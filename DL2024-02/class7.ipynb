# Calculo Matricial
Asumamos que $A$ es un matriz real $m \times n$, donde $a_{ij}=a_{ij}(x)$
donde $x \in \mathbb{R}^p$.  Mas concretamente, definimos la derivda parcial de $A$ con respecto a  la componente $k$ del vector $x$ como

$$ \frac{\partial A}{\partial x_k} = \left [
    \frac{\partial a_{ij}}{\partial x_k} \right ], \quad , \quad k=1,2, \cdots, p \quad ; \quad i=1,2, \cdots, m \quad , \quad j=1,2, \cdots, n.$$

Un ejemplo

Sea
$$A = \begin{pmatrix} x^2 y & x - y \\ \sin (xy) & \exp(x^2 + y^2) \end{pmatrix}  $$

$$ \frac{\partial A}{\partial x} =
\begin{pmatrix}
2 x y & 1 \\ y \cos(xy) & 2 x \exp(x^2 + y^2)
\end{pmatrix}
$$


**Teorema 1.**: Si $A$ es una matriz constante (ninguna componente depende de $x$, entonces $\partial A/\partial x_k=0$

**Prueba**
$$ \frac{\partial A}{\partial x_k} = \left [
    \frac{\partial a_{ij}}{\partial x_k} \right ]= [ 0 ] $$

Recuerden de calculo que $d ax/x=a$. De la misma forma   

**Teorema 2.**
$$
\frac{\partial (Ax)}{\partial x_k} = \nabla_k(A x) = A_k
 $$
donde $A_k$ es la $k$ columna de $A$.


**Prueba**
Sea $Ax = y$,  donde $A$ es una matriz constante

$$y_i = \sum_{j=1}^n a_{ij} x_j $$
de forma que

$$ \frac{\partial y_i}{\partial x_k} = \sum_{j=1}^n a_{ij} \frac{\partial x_j}{\partial x_k}   .$$

Ahora
$$\frac{\partial x_j}{\partial x_k} = \delta_{kj} .$$

$$ \frac{\partial y_i}{\partial x_k} = \sum_{j=1}^n a_{ij} \frac{\partial x_j}{\partial x_k} = \sum_{j=1}^n a_{ij} \delta_{kj} = a_{ik}  .$$
dode $a_{ik}$ para $i=1,2, \cdots, m$ es la columna $k$ de $A$

De alguna forma podemos decir que $\partial Ax/\partial x = A$.

**Teorema 3**: Asuma $f(x), g(x)$ funciones vectoriales de $x \in \mathbb{R}^n$ entonces


$$ \frac{\partial  \langle  f , g \rangle }{\partial x_i} =
\left \langle \frac{\partial f}{\partial x_i}, g   \right \rangle +
\left \langle f,  \frac{\partial g}{\partial x_i }   \right \rangle +
  .$$

**Prueba**

$$\frac{\partial \langle f, g \rangle}{\partial x_i}= \frac{\partial}{\partial x_i} \sum_{j=1}^n \overline{f_j(x)} g_j(x) =\sum_{j=1}^n \frac{\partial \overline{f_j(x)}}{\partial x_i} g_j(x) + \overline{f_j(x)} \frac{\partial g_j(x)}{\partial x_i} = \left \langle \frac{\partial f}{\partial x_i}, g   \right \rangle +
\left \langle f,  \frac{\partial g}{\partial x_i }   \right \rangle  .$$










De alguna forma esto prueba que

$$\nabla \langle f , g \rangle  = \langle \nabla f, g \rangle + \langle f, \nabla g \rangle .$$
Pero toca mirarlo con cuidado. Por que $f,g$ son vectores, pero
 $\nabla f$ y $\nabla g$ son matrices.
 Realmente lo que debemos decir es que la componete $k$ del gradiente de $f$ es la columna $k$ de la matriz

$$\nabla_k \langle f , g \rangle  = \langle \nabla_k f, g \rangle + \langle f, \nabla_k g \rangle .$$

**Ejemplo 1**: Sea $f(x)=\frac12 \| A x - b \|^2 $ con $x \in \mathbb{R}^n$ y $A$ es una matriz real $m \times n$. Encuentre $\nabla f$.

**Solucion:**
Escribimos $f$ como un producto interno

$$f(x) = \frac12 \langle Ax - b, Ax - b \rangle .$$
Usamos la notacion $\nabla_k f= (\nabla f)_k$

Del teorema anterior

\begin{eqnarray}
\nabla_k f(x) &=& \frac12 \left [ \langle \nabla_k(Ax - b), (Ax - b) \rangle   +
\langle Ax - b , \nabla_k(Ax - b) \rangle \right ]  \\
&=& \langle \nabla_k (Ax -b), Ax-b \rangle  \quad , \text{producto interno en reales conmuta} \\
&=& \langle A_k, Ax - b   \rangle \quad , k=1,2, \cdots, n
\end{eqnarray}

Quitando la $k$ y usando la definicion de producto interno.

$$\nabla f(x) = A^*(Ax - b) .$$


A esta ecuacion se le llama las ecuaciones normales.

**Motivacion**:
Si queremos resolver $Ax=b$, y $A$ ni siquiera es cuadrada, no tiene solucion, un "truco" que se puede hacer es multiplicar por la adjunta a ambos lados

$$A^* A x = A^* b .$$
Si $\nabla f(x) = 0$ (esta la condicion de maximos/minimos


$$\nabla f(x) = A^*(Ax - b) = 0 .$$
entones

$$ A^* A x = A^* b .$$
Si $A^* A$ tiene inversa entonces

$$x = (A^* A)^{-1} A^* x .$$
Esta es la solucion de **minimos cuadrados ordinarios**.











## Minimos cuadrados regularizados.

Asumimos $A \in \mathbb{R}^{m \times n}$, $b \in \mathbb{R}^m$, $x \in \mathbb{R}^n$

El problema central es $Ax=b$. Queremos hallar $x$.

* Si $A$ tiene inversa, $x = A^{-1} b$.
Sin embargo estadicamente la probabilidad de que $A$ tenga inversa es 0.
Que pasa si $A$ es singular.

Hay dos posibilidades:

1. La matriz a esta **sobre determinada** (overdetermined). El sistema $Ax=b$ tiene mas ecuaciones independientes que incognitas.
El espacio nulo es 0. $\mathcal{N}(A) = \{ 0 \}$

Ej:

$$ 3 x + 2y = 5 \\ 4 x -4y = 6 \\ 5x + 4y = 2 \\ 3x - 3y = 0 $$

2. La matriz a este **sub determinada** (underdetermined). En esta caso hay mas incognitas que ecuaciones
El espacio nulo no es el trivial 0, $\mathcal{N}(A) \ne \{ 0 \}$.

$$ 3x + 2y = 5 .$$
Tiene un numero infinito de soluciones $x=1, y=1$, $y=0,x=5/3$. Es una linea de soluciones.



El problema de minimos cuadrados (ordinarios) se refiere a la minimizacion
de

$$ f(x) = \| Ax - b \|^2  . $$


Puede pasar que $A^* A$ tenga autovalores  muy pequenhos (valores singulares). Eso genera un "condition number" muy grande e inestabilidad.
No es suficiente la funcion objetivo (en Ingles: Target function, Cost function , loss function) no es suficiente cuando los valores singulares se acercan a 0. Por que entonces no existe la inversa de
$A^* A$.
Hay un teorema que dice que si las columnas de $A$ no son **linealmente independientes** entonces $A^*A$ no tiene inversa.

LLego un senhor (Ruso) Tikhonov, amplio la funcion objetivo


$$ f(x) = \|  A x - b \|^2 + \| \Gamma x \|^2 .$$
donde $\Gamma$ es una funcion positiva definida.




Con esta funcion objetivo, y usando la teoria que se desarrollo en esta clase (hallar el gradiente $\nabla f = 0$) se encuentra que

$$x = ( A^* A + \Gamma^* \Gamma)^{-1} A^* b .$$

Esta ecuacion corresponde a minimos cuadrados **regularizados**


Es comun que se use un $\Gamma = \sqrt{\lambda} I$
resulta

$$ x  = (A^* A + \lambda I)^{-1} A^* b $$
ojo $\lambda > 0$ y ojala pequenho.
A este tipo de regularizacion se le llama **ridge regression**.

Podemos llegar a esta formula con un metodo heuristico.

Pensemos en que queremos resolver $Ax = b$.

* Si $A$ tiene inversa $x= A^{-1} b$ y listo.
* Si $A$ no tiene inversa, el proximo paso es construir las ecuaciones normales (multiplicando por $A^*$ a ambos lados, es decir
$$ A^* A x = A^* b .$$
Si $A^* A$ tiene inversa  entonces $x = (A^* A)^{-1} A^* b$ listo.
Pero Si $A^* A$ no tiene inversa o  **ojo** es inestable (un numero de condicion grande, por decir algo que $\sigma=10^{-8}$.
Toca regularizar,

Como se regulariza, como los autovalores de $A^*A$ son cero o muy pequenhos, los empujamos hacia arriba sumandole $\lambda I$,
y entonces buscamos

$$ (A^* A + \lambda I) x = A^* b .$$

De aca

$$x = (A^* A + \lambda I)^{-1} A^* b \tag{2} .$$

La funcion objetivo que me genera esta solucion de minimos cuadrados regularizados es


$$J(x) = \frac12 \| Ax - b \|^2 + \lambda \| x \|^2 .$$
La solucion (2) se obtiene de hacer $\nabla J(x) = 0$

Otro enfoque del problema es a traves **multiplicadores de Lagrange**.
Es lo mismo que maximos/mininos con restricciones.

El problema se escribe, hallar $x$ tal que


\begin{eqnarray}
\min_x \frac12 \| Ax - b \|^2 \\
\text{sujeto a la restriccion} \quad \| x \|^2 \le c
\end{eqnarray}
