## Tensores.
Estrictamente hablando existen los campos de algebra tensorial y calculo tensorial. Aca, tocamos, por encima, el algebra tensorial. En un apendice de las notas hago una resenha de calculo tensorial. El area que reune estos dos se llama **analisis tensorial**. Pero aca vamos ver por encima conceptos muy basicos.

Desde vista fisico un **tensor** es un obtjeto invariante ante cambios de coordenadas y por las leyes fisicas se formulan en terminos de tensores.

**Motivacion**: Piense en un conjunto de imagenes (las etiqueta con el indice $i$), pero cada imagen tiene pixeles en dos dimensiones $j,k$ (horizontal y vertical) y cada pixel podria tener 3 (RGB)  (o 4 CYMB), aca tendriamos otro indice $l$. Este objeto no se estudia "normalente" en algebra lineal. Este objeto tiene 4 indices. En la teoria tensorial a este objeto se le llama un tensor de rango 4. **Ojo**: En la ciencia de la computacion (ML) a esto se le llama **dimension** pero no es asi.

Por ejemplo $a_{ijkl}$, $i,j,k,l=1,2,4$, las dimensiones (desde el punto de vista del Algebra lineal) son $4^4=256$. Pero para `NumPy` este "objeto" tiene 4 dimensiones.

Para nosotros no complicarnos la vida para ML, un tensor es un objeto multi-indice. Sin embargo voy a motivar la introduccion de tensores con formas lineales.

La teoria de la relatividad se alimento de los tensores. Ellos ayudaron mucho.

## Tensores para matematicos.
Transformaciones multi-lineales. (una generalizaion del algebra lineal)

Un poco de teoria :(
Veamos.

* Asumamos $x \in \mathbb{R}$, $y \in \mathbb{R}$, $a \in \mathbb{R}$.  La transformacion **lineal**

$$ Tx = y = ax $$
Esta es una linea recta de pendiente $a$. Aca $y$ es un tensor de rango 0 (es un escalar).

* Ahora pensemos que $x \in \mathbb{R}^n$, $y \in \mathbb{R}^n$ y queremos transformar

$$ Tx = y = a x .$$
Quien es $a$? Si estudia el algebra lineal $a$ es una matrix $n \times n$. Como se obtiene esta matriz?

Se elige una base

$$ e_1 =
\begin{pmatrix} 1 \\ 0 \\ \vdots \\ \vdots \\ \vdots \\  0  \end{pmatrix}
\quad , \quad
e_2 = \begin{pmatrix} 0 \\ 1 \\ \vdots \\ \vdots \\ \vdots \\  0  \end{pmatrix}
\quad , \quad
\cdots  \quad ,
e_n = \begin{pmatrix} 0 \\ 0 \\ \vdots \\ \vdots \\ \vdots \\  1  \end{pmatrix}
\quad , \quad
.$$
Del Algebra lineal


$$x = \sum_{i=1}^n x_i e_i.$$

$$Tx = T \left ( \sum_{i=1} x_i e_i  \right ) = \sum_{i=1}^n x_i T(e_i).$$
$T(e_i) \in \mathbb{R}^n$
Usando algebra lineal $Tx$ es


Del algebra lineal el operador $a$

$$ a =
\begin{pmatrix}
| & | & \vdots & \vdots & | \\
T(e_1) & T(e_2) & \vdots & \vdots & T(e_n)  \\
| & | & \vdots & \vdots & |  
\end{pmatrix}
$$




En esta transformacion lineal $y$ es un tensor de rango uno. Es lo que nosotros conocemos como vector.

*  $x \in \mathbb{R}^{n \times m} $, $y \in \mathbb{R}^{p \times q}$

Queremos $y = ax$
Quien es $a$?

Simpre en el algebra lineal la transformacion $T$ ($a$), se encuentra escribiendo el vector $x$ en terminos de las bases.

Quien es la base del espacio $\mathbb{R}^{n \times m}?$.

Empecemos bajito.  $n=m=2$.

Los "vectores" base.


$$
e_{11} = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}  \quad , \quad
e_{12} = \begin{pmatrix} 0 & 1 \\ 0 & 0 \end{pmatrix}  \quad , \quad
e_{13} = \begin{pmatrix} 0 & 0 \\ 1 & 0 \end{pmatrix}  \quad , \quad
e_{14} = \begin{pmatrix} 0 & 0 \\ 0 & 1 \end{pmatrix}  \
.$$









Ejemplo:

Generalizo. Para un espacio de $n \times m$ dimensiones.



image.png

De la mima forma que en la matrix $abcd$ de arriba ($2 \times 2$), tenemos que si $x \in \mathbb{R}^{m \times n}$


$$ x = \sum_{i=1}^n \sum_{j=1}^m x_{ij} e_{ij}.$$
y
$$ T(x) = \sum_{i=1}^n \sum_{j=1}^m x_{ij} T({e_{ij}}) .$$
cada $T(e_{ij}) \in \mathbb{R}^{p \times q}$

Podemos indicar $T$ con los indies $s=1, \cdots p$, $t=1, 2, \cdots, q$ y escribir

$$ T_{st}(x) = \sum_{i=1}^n \sum_{j=1}^m x_{ij} T_{st}({e_{ij}}) . \tag{1} $$

El objeto $a$ es de la forma $T_{st}(e_{ij})$.

$$ a_{st}^{ij} = T_{st}(e_{ij})$$.
y $y$ es un tensor de rango 2, dimensiones $n\times m$.

Como las sumas son muchas, entonces se suprimen basadas en el siguiente principio. En la ecuacion (1), los indices $i,j$ estan repetidos (cada uno dos veces). A estos indices se le llaman **dummies**, mientras que los otros indices solo aparecen una vez. Son **libres** (free). Einstein, en publicacion de la teoria de la relatividad elimino las sumas y se asume que la suma se hace sobre los indeices **dummies**.  La expresion (1) se escribe



$$ T_{st}(x) = x_{ij} T_{st}({e_{ij}}) . \tag{2} $$

Podemos escrbir

$$ T_{st}(x) = x_{ij} T_{st}(e_{ij}) = a_{st}^{ij} x_{ij} .$$

Generalicemos.


Asumamos
$x \in \mathbb{R}^{n_1 \times n_2 \times \cdots \times n_j}$
$y \in \mathbb{R}^{m_1 \times m_2 \times \cdots \times m_j}$
Los vectores base son de la forma
$e_{p_1 p_2 \cdots p_j}$ tienen $1$ en la posicion $p_1 p_2 \cdots p_j$ y ceros en el resto. Entonces podemos escribir


$$x = x_{p_1 p_2 \cdots p_j} e_{p_1 p_2 \cdots p_j} .$$
Todos los indices aca son **dummy**

y

$$ T(x) = x_{p_1 p_2 \cdots p_j} T(e_{p_1 p_2 \cdots p_j}).$$


Donde vide $T$? $T \in \mathbb{R}^{m_1 \times m_2 \times \cdots m_j} $ De forma que podemos escribir


$$ \left [ T(e_{p_1 p_2 \cdots p_j})   \right ]_{q_1 q_2 \cdots q_k} =  a_{q_1 q_2 \cdots q_k}^{p_1 p_2 \cdots p_j}$$

Al final lo que concluimos

$$ T_{q_1 q_2 \cdots q_k}(x) = a_{q_1 q_2 \cdots q_k}^{p_1 p_2 \cdots p_j} x_{p_1 p_2 \cdots p_j} .$$

La cantidad $T_{q_1 q_2 \cdots q_k}(x)$ es un  **tensor de rango $k$**

Otra forma de pensar, tal vez simple es. Asuma que

$$x \in \mathbb{R}^{n_1 \times n_2 \times \cdots \times n_j} $$











Podemos pensar que

$$x = (x_1, x_2, \cdots , x_j) .$$
donde
$x_1 \in \mathbb{R}^{n_1}$,
$x_2 \in \mathbb{R}^{n_2}$,
$\cdots$
$x_k \in \mathbb{R}^{n_k}$,
donde $k=1,2, \cdots, j$.

El operador $T$ es lineal en cada una de las variables.

### Ejemplo con `NumPy`

```python
# ejemplo con numpy
import numpy as np
x = np.arange(24).reshape(2,3,4)
print(x)
```

```python
# extraemos toda la informacion de las dimensiones de el objeto x
print("la longitud del vector (len)", len(x)) # sobre la primera dimension
print("el rango del tensor es ", np.ndim(x))
print("la forma (shape) del tensor es", np.shape(x))
print("el tamanho del pector es", np.size(x))
```
