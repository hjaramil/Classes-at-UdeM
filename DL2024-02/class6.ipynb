Describamos el patron. Asumimos un kernel $M_y \times N_y$
y los datos $M_x \times N_x$, desarrollamos (unroll,estiramos,vectorizamos, flatten) ambas matrices $y_i, i=0,1, ,M_y \times M_x$, $x_i, i=0,1, \cdots M_x \times N_x$.


Necesitamos crear una matriz $y$ como sigue

* La primera fila. $y_0, y_1, \cdots, y_{N_y-1}$, despues
$N_x - N_z$ ceros, despues $y_{N_y}, \cdots y_{2 N_y-1}$, y luego $N_x - N_y$ ceros, y asi hasta que se estire toda la matriz $y$. La parte de esta fila $y_{(M_y-1)N_y},\cdots ,y_{M_y N_y} $ y  $N_x - N_y$ ceros.

* Las filas posteriores. Usted corre $y$ a la derecha, hasta que alcance el borde derecho, y aca en cada fila se agrega un 0 al principio y se corre todo un lugar y quita un cero al final.
Una vez llega al final, toca saltar de fila en la "figurita"
toca agregar $N_x$ ceros al inicio y correr lo que ya se tiene $N_x$ lugares y trunca al final $N_x$ (ceros?)

A este tipo de **upscaling** se le llama **convolucion transpuesta**.  En este momento debemos recorbrar la forma original de los datos (tensores de rango 2). Vamos a  ver como se hace el problema con las matrices originales.

$$ x = \begin{pmatrix}  
x_{00} & x_{01} & x_{02} & x_{03} \\
x_{10} & x_{11} & x_{12} & x_{13} \\
x_{20} & x_{21} & x_{22} & x_{23} \\
x_{30} & x_{31} & x_{32} & x_{33} \\
\end{pmatrix}  \quad , \quad y=
\begin{pmatrix}
y_{00} & y_{01} & y_{02} \\
y_{10} & y_{11} & y_{12} \\
y_{20} & y_{21} & y_{22} \\
\end{pmatrix}
\quad , \quad
z =
\begin{pmatrix}
z_{00} & z_{01} \\
z_{10} & z_{11}
\end{pmatrix}
.$$

Por observacion en los arreglos de $y$ estan reversados. Es como si convolvieramos $z$ con un filtro $y$ (reversado, de ahi tambien la palabra "transposed convolution")
Ese filtro es

$$ y =\begin{pmatrix}  y_{22} & y_{21} & y_{20} \\
y_{12} & y_{11} & y_{10} \\ y_{02} & y_{01} & y_{00} \end{pmatrix}  $$

Conteo.

En problema que acabmos de ver, El filtro es $3 \times 3$,
el padding es $p=2$, los datos de entrada son $2 \times 2$,
el resultado es

$$(M_z - M_y + 2 p + 1) \times (N_z - N_y + 2p + 1)= 4 \times 4 .$$

En resumen.

* Para la convolucion directa, se da un arreglo 2D $x$ con dimensiones $M_x, M_y$, un filtro $y$ con dimensiones $M_y, N_y$, podemos hacer la convolucion sin padding $p=0$, y no striding $s=1$. La salida tiene dimensiones $M_z=M_x - M_y + 1$, $N_z = N_x - N_y + 1$.

* **convolucion transpuesta**: La entrada es un arreglo $z$ de dimensiones $M_z, N_z$ y un filtro con dimensiones $M_y, N_y$. Debemos reversar el orden del filtro $y$ (original). Una forma sencilla, se hace "flatten" y se reversa el resultado, luego le hacemos "reshape". Se hace padding $p$ en los datos ($z$), que numero es $p$?

Veamos

$$ M_z - M_y + 2 p_M + 1 = M_x  \quad , \quad
 N_z - N_y + 2 p_N + 1 = N_x  $$

La respuesta a la pregunta de cuanto padding hacer es, despeje.

$$ p_M = \frac{M_x - M_z + M_y - 1}{2}.$$
$$ p_N = \frac{N_x - N_z + N_y - 1}{2}.$$
En el ejemplo de esta clase
($M_x=N_z, M_y=N_y, M_z$)

$$N_x=4, N_y=3, N_z=2 .$$
entonces
$$ p_N= p_M =  \frac{4 - 2 + 3 - 1}{2} = \frac{4}{2}=2.$$
