## Historia:

La historia del DL esta pegada de la ANN y del ML.

DL es parte de ML. $\text{DL} \subset \text{ML} \subset \text{AI} $.

* **1960s** :Henry J. Kelly desarrollo el algoritmo de "back-propagation".  Stuart Dreyfus desarrolo en 1962 unsa version mas simple de propagacion hacia atras.
* **1970s** : Seppo Linnainmaa escribio su tesis de maestria en FORTRAN en "back propagation".
Kunihiko Fukushima: Invento las CNN en 1979. Importantisimas.
* **1980s** : Yann LeCun (jefe de investigacion Facebook).
El administra el MNIST. En 1989 implemento el back propagation para resolver el problema del MNIST (lo veremos en clase).
Para mi, fue el primero que le dio poder al back propagation.
Yann Lecun trabajo con CNN.
* **1990s**: Sepp Hochreiter invento los LSTM: Long Short Term Memory. Redes neuronales para secuencias, muy importantes para NLP: Natural Language Processing. Predecir sequencias (Time Series).  En 1999 aparecieron los GPU: Graphic Processor Unit.
Esto aceleron el desarrollo del DL.
* **2000s**: Vanishing Gradient Problem. Cuando se hace backpropagation el gradiente podria hacerse tan pequenho que no habria mas aprendizaje. $\theta_{k+1} = \theta_{k} - \eta \nabla J$ si $\| \nabla_{\theta} J \| \ll 1$ entonces
$\theta_{k+1} \approx \theta_k$. De otra forma tambien existe el **exploding gradient**. El gradiente se va para infinito. En la clase veremos ejemplos de esto.
En el 2009, la profesora (China) Fei-Fei Li creo el program
[imagenet](https://www.image-net.org/). ES una base de datos co mas de 15 millones y 22 mil clases.
* **2010s**: En el 2011 se creo un **reto** con imagenes de alta resolucion con 1000 clases. De ese reto salieron muchas tecnologias importantes, de las cuales, varias las veremos en este curso. (por ejemplo el AlexNet y otras)
Se creo el program ImageNet Large Scale Visual Recognition Challenge (ILSVRC). En el 2012 Google Brain libero el proyect **The Cat Experiment**, fundado por
Andrew Ng. Aca se exploran las dificultades del ML no supervisado. Este proyecto se hizo sobre una red de mas de 1000 computadores.
* En el 2010 comenzo la competicion de **Kaggle**.
Esta comunidad impulso mucho el desarrollo de DL.

Es importante entender la diferencia entre "library" y "framework.
[Este sitio](https://www.freecodecamp.org/news/the-difference-between-a-framework-and-a-library-bd133054023f/) ilustra muy bien estas diferencias.


"
When you use a library, you are in charge of the flow of the application. You are choosing when and where to call the library. When you use a framework, the framework is in charge of the flow. It provides some places for you to plug in your code, but it calls the code you plugged in as needed."

### Notas sobre el lenguaje.
* Tensores: no son simplemente objetos multi-indice como pretende definirlos la comunidad de AI. Son objetos invariantes ante cualquier sistema coordenado. Pero, bueno, vamos a pegarnos a la comunidad de AI.

* La palabra multidimensional esta mal usada en la comunidad AI.
Ellos dicen que un tensor de dimenion 3 es $a_{ijk}$. Pero
eso no es asi. La dimension del tensor $a_{ijk}$, $i,j,k=1,2,3$
es $3 \times 3 \times 3 = 27$. (quienes son ellos `NumPy`).
El vector $x=(a,b,c)$ es un vector de dimension 3 y de rango 1.
$a_{ijk}$ es de dimension 27 de rango 3.

* La palabra convolucion que se usa en Computer Science (CS), es realmente correlacion. Pero, bueno, nosotros nos pegamos a ese error.

* La palabra deconvolucion, tambien esta mal usada, deber ser convolucion transpuesta.
