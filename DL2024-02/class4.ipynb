# Correlacion en tensores de orden superior (rango 2 o mas)
Repitiendo el analisis de la clase anterior, pero ahora para tensores de rango 2 (funciones de dos variable).


$$h(\tau , \rho) = \int_{D_1} \int_{D_2} f(t,r) g(t + \tau, r + \rho) dt dr .$$

En la computacion usamos el discreto.


$$ z_{jk} = (x \star y)_{jk} = \sum_m \sum_n y_{mn} x_{m+ j \; n+k}  .$$

Para definir los limites de las sumas, asumamos que los dator $x$, tienen
dimensiones $(M_x, N_x)$ y que el filtro tiene dimensiones $(M_y, N_y)$.
Ojo, generalmente el filtro es mas pequenho que los datos. Es decir
$M_y < M_x$, $N_y < N_x$. Si son iguales y no hay padding lo que tiene es una muestra de salida, el producto punto $x \cdot y$.
Como necesitamos

$$ m+j < M_x  \quad , \quad n+k < N_x .$$
entonces escribimos


$$ z_{jk} = (x \star y)_{jk} = \sum_{m=0}^{M_y-1} \sum_{n=0}^{N_y-1} y_{mn} x_{(m+j)   \; (n+k)} .$$
$$ j=0,1, \cdots, M_x - M_y + 1  \quad , \quad k = 0, 1, \cdots, N_x - N_y  .$$


El tamanho de la salida es

$$\#(z_{ij}) = (M_x - M_y + 1) \times (N_x - N_y + 1).$$


**Agregamos padding**
el padding es en las dos direcciones. Generalmente el padding es el mismo en cada direccion.  La formula de terminos de salida con padding.


$$\#(z_{ij}) = (M_x - M_y + 2 p_M + 1) \times (N_x - N_y  + 2 p_N + 1).$$


Si queremos que la salida tenga el mismo numero de muestraas que la entrada, entonces. Con padding la formula es

$$ p_M   = \frac{M_x -1}{2} \quad , \quad p_N = \frac{N_x-1}{2}.$$
Esto lo llaman en `TensorFlow`:  **same**

Para definir los limites de las sumas, asumamos que los dator $x$, tienen
dimensiones $(M_x, N_x,O_x)$ y que el filtro tiene dimensiones $(M_y, N_y,O_y)$.
Ojo, generalmente el filtro es mas pequenho que los datos. Es decir
$M_y < M_x$, $N_y < N_x$, $O_y < N_x$. Si son iguales y no hay padding lo que tiene es una muestra de salida, el producto punto $x \cdot y$.
Como necesitamos

$$ m+j < M_x  \quad , \quad n+k < N_x  \quad , \quad . o+l < O_y$$
entonces escribimos


$$ z_{jk} = (x \star y)_{jk} = \sum_{m=0}^{M_y-1} \sum_{n=0}^{N_y-1} \sum_{l=0}^{O_y -1} y_{mn} x_{m+j , n+k, o + l} .$$
$$ j=0,1, \cdots, M_x - M_y + 1  \quad , \quad k = 0, 1, \cdots, N_x - N_y  \quad , \quad l = 0, 1, \cdots, O_x - O_y .$$


El tamanho de la salida es

$$\#(z_{ijl}) = (M_x - M_y + 1) \times (N_x - N_y + 1) \times (O_x - O_y + 2 p_O  + 1).$$


**Agregamos padding**
el padding es en las dos direcciones. Generalmente el padding es el mismo en cada direccion.  La formula de terminos de salida con padding.


$$\#(z_{ijl}) = (M_x - M_y + 2 p_M + 1) \times
(N_x - N_y  + 2 p_N + 1) \times
(O_x - O_y  + 2 p_O + 1)$$


Si queremos que la salida tenga el mismo numero de muestraas que la entrada, entonces. Con padding la formula es

$$ p_M   = \frac{M_x -1}{2} \quad , \quad p_N = \frac{N_x-1}{2}
 \quad , \quad p_0 = \frac{O_x-1}{2}.$$
Esto lo llaman en `TensorFlow`:  **same**

**Strides**:

$$ z_{jk} = (x \star y)_{jk} = \sum_{m=0}^{M_y-1} \sum_{n=0}^{N_y-1} \sum_{l=0}^{O_y -1} y_{mn} x_{m+j , n+k, o + l} .$$
$$ j=0,s_x,  2 s_x, \cdots \le  M_x - M_y + 1  \quad , \quad k = 0, s_y, 2 s_y  \cdots  \le N_x - N_y  \quad , \quad l = 0, s_z, 2 s_z  \cdots, \le O_x - O_y .$$
donde $s_x, s_y s_z > 1$. Que ganancia hay? que el objeto de salida se reduce en $s_x \times s_y \times s_z$ dimensiones.
