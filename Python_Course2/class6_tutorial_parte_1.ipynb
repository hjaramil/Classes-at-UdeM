## Operaciones con cadenas de caracteres
### seleccion de campos
Syntax. Let ```a``` be the string

``` a[i:j] ```  
Takes from indices ```i``` to ```j``` not included ```[i, j)``` where
```i``` is the start index and
```j-1``` the ending index

```a[i:]  ```  goes frim i to the end
```a[:i] ``` goes from begining to ```i-1```

````python
a[2:7]
````

````python
print(a[0])
print(a[1])
print(a[2])
````

````python
a[6]
````

````python
a[:3]
````

````python
a[3:]
````
