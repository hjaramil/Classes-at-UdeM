## Diferencias entre listas y arreglos.
Que es un arreglo (tensor, de tensores hablamos luego). Es una coleccion de numeros homogenamente distribuidos en capas, filas, columnas, etc.



````python
L=[1,3,5,7]
A=np.array(L)

print("lista", L)
print("Arreglo", A)
````

````python
# usted le puede agregar un elemento a una lista
L.append(6)
print(L)
````

````python
A.append(6)
````

````python
L=[1,3,5,7]
A=np.array(L)


B=2*A
M=2*L

print(f"B={B}")
print(f"M={M}")
````

````python
C=M**2
````

````python
C=A**2
print(C)
````

````python
np.sin(L)
````

````python
np.sin(A)
````

En resumen:

* En la impresion, la lista la imprime con comas y el arreglo con espacios. No hay puntos ni a favor ni en contra. Sintaxis

* Las listas tienen funciones que no tienen los arreglos. Por ejemplo, ```append```, ```pop``` y demas . Punto a favor de las listas.

* En la multiplicacion de un entero por una lista, se repite la lista, tantas veces como el entero. En la multiplicacion de un entero por un arreglo , cada entrada del arreglo se multiplica  por el entero. 

* Toda el algebra lineal, se debe hacer sobre arreglos, no sobre listas.
Aunque muchas funciones aplicadas a una lista dan los mismos resultados que un arreglo. Por ejemplo ```np.sin(L)``` y ```np.sin(A)```

# Graficas en ```Python```:

````python
import matplotlib, numpy
from matplotlib.pyplot import plot, xlabel, ylabel, title, text, \
    axis, grid, show, legend

x=np.arange(0,2*np.pi,0.01)
s=4
title(f"$cos {s} x$")
plot(x, np.cos(s*x))
plot(x,x)
show()
title(f"sin{s} x")

plot(x,np.sin(s*x))
show()
````

````python
title("$\sqrt{ 4 x}$")
var = 123

xlabel('$x$')
ylabel('$\sqrt{4 x}$')
text(1, 4,f'text$_{{{var}}}$')
grid(True)
plot(x,np.sqrt(s*x), label="$\sqrt{4 x}$")
plot(x, x**4/4, label="$x^4/4$")
legend()

show()
````
