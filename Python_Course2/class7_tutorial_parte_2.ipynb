# Operadores entre bits
La representacion binaria de $5$ es $101$, la representacion binaria de $3$ es $011$. La operacion $|$ compabara bit por bit y solo si los dos bit son 0 arroja 0 (or, disjuncion "o"), de otra forma arroja 1.

Es decir:

\begin{equation}
101  \\
\, 011  \\
----  \\
111
\end{equation}
de forma que "$5 | 3 = 7$" (bitwise or).

El "o" exclusivo "^" solo es 0 cuando ambos son iguales y uno cuando son diferentes

\begin{equation}\
101  \\
\, 011  \\
----  \\
110
\end{equation}

La conjuncion "&". Solo cuando los dos son verdaero.

\begin{equation}\
101  \\
\, 011  \\
----  \\
001
\end{equation}

```python
5 | 3
```

```python
5^3
```

```python
5 & 3
```

### Operadores de desplazamiento (shift).
El operador ">>" desplaza los bits a la derecha. Por ejemplo
5>>2, que hace?

$$ 101 >>2 \rightarrow 001$$

El operador "<<" desplaza los bits a la izquierda. Por ejemplo
5<< 2 que hace?

$$ 101 <<2 \rightarrow  (10100)_2 = 20$$

```python
5>>2
```

```python
5<<2
```

## Conjuntos
separados por comas y entre llaves



```python
L=[10,1,3,4,4,1,"a", "hola"]
print("L=", L, "es de tipo", type(L))
setL = set(L)  # casting de lista a conjunto
print("el conjunto setL", setL, "es de tipo", type(setL))
```

```python
M = {10, 1, 3, 4, 4, 1, 'a', 'hola'}
print(M)
```

```python
### Operaciones entre conjuntos
A = set(['a', 'b', 'c'])
B = {'b', 'c', 'd'}

print("A tiene", len(A), "elementos")
```

```python
print("Esta 'a' en A?", 'a' in A)  # pertenencia
```

```python
print("A unido B", A.union(B))
```

```python
print("A intersectado B", A.intersection(B))
```

```python
print("A \ B", A.difference(B))
```

```python
print("A superconjunto de B", A.issuperset(B))
```

```python
print("A subconjunto de B", A.issubset(B))
```

```python
print("A diferencia simetrica con B", A.symmetric_difference(B))
```
