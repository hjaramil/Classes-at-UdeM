## funcion ```map```

Syntaxis:
```
map(funcion, objeto)
```
La funcion se aplica al objeto.

````python
def a_mayuscula(s):
    return str(s).upper()

cadena='abc'
map_cadena = map(a_mayuscula, cadena)
````

````python
print(list(map_cadena))
````

````python
def cuad(i):
    return i**2

L=[1,3,5]
print(list(map(cuad, L)))
````

## Funciones "lambda"
La sintaxis:

```
lambda variable  : retorno
```



````python
f = lambda x: x**2
f(3)
````

````python
g = lambda x: x**3 - 3*x + 5
g(5)
````

````python
# equivalente a
def pol(x): return x**3 - 3*x + 5

pol(5)
````

Diferencias entre funciones lambda y funciones regulares:

* La funcion regular tiene ```def```. La funcion lambda tiene un ```lambda```
* La funcion tiene un  ```return``` explicito. El ```return``` en la funcion lambda no existe.
* El nombre de la funcion esta despues de ```def``` en las funciones regulares. En las funciones ```lambda```. El nombre se asigna a una variable a la izquierda. O no se asigna. Veamos un ejemplo donde se usan funciones lambda "anonimas". Las funciones lambda son ideales para instrucciones de una sola linea. Instrucciones cortas.

````python
f = lambda x: x*x
p=f(3)
print(p)
````

````python
print("el tipo de la variable f es", type(f))
````

````python
g=f
g(2)
````

````python
# tambien con funciones regulares se puede hacer esto
g=pol
g(5)
````

## Filtros

Sintaxis:
```
filter(expresion logica, una iterable)
```

````python
impares = filter( lambda i : i%2 ==1 , range(10))
list(impares)  # imprime el valor de la lista range(10) que es verdadero para la condicion. 
````

````python
impares = map (lambda i : i%2 ==1 , range(10))
list(impares)
````

````python
A = {1,2,3,4,5}
impares = filter( lambda i : i%2 , A)  # elige el valor del conjunto A, que satisfaga la condicion
set(impares) 
````

````python
# no es necesario que sea una funcion lambda
def imp(i):
    return i%2==1

# necesito la palabra "lambda" para poder fijar la variable independiente "i"
impares = filter(lambda i: imp(i) , A)  # elige el valor del conjunto A, que satisfaga la condicion
list(impares)
````
