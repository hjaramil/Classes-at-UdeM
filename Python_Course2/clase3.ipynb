# Los numeros mas grandes y mas pequenos de la precision doble (64 bits IEEE).

El **epsilon** de la maquina: Es la distancia entre $1$ y el siguiente numero.
$1+\epsilon$ es el siguiente numero despues de $1$ y $1-\epsilon$ es el anterior
y el conjunto $(1-\epsilon, 1+\epsilon)=1$. 

Ese numero es $2^{-52}$. Corresponde a la mantisas masa pequena.

$$ 2^{-52} = 2.220446049250313e-16$$.

El siguiente algoritmo se usa para encontrar el
epsilon de la maquina.

```python
eps=1
while eps+1 != 1:
    trueps = eps
    eps /= 2

print(trueps)
```

Cual es el numero mas grande del computador?
Cual es mantisa ($f$) mas grande

$$.11111 \cdots 1$$.
Como calculamos este numero.

\begin{eqnarray}
(.1111 \cdots 1)_2 = \left ( \frac12 \right )^1 + \cdots\left ( \frac12 \right )^2 + \cdots \left ( \frac12 \right )^{52}.
\end{eqnarray}
Sea $x=1/2$

$$ x + x^2 + \cdots + x^52 = x(1 + x + x^2 + \cdots x^{51}) = x \frac{1 - x^{52}}{1-x} = 1 - x^{52} $$
La mantisa mas grande es

$$ 1 - 2^{52} =0.9999999999999998 $$

El exponente mas grande es
$$(11111111111)-1023 = 1024$$

## Underfloat, Overfloat
Tenemos la flexibilidad en Python de declarar variables de distinta precision (numero de bits).

* precision entero (short or half sin signo) 16 bits, formato "h".
* prercision entero (short or half con signo) 16 bits, formato "H"
* entero de 32 bits (sin signo) en formato "i"
* entero de 32 bits (con signo) ene formato "I"
* numeros de punto flotante.

Ejemplos en Python

```python
# signed half
from array import array

a = array('h')  # 16 bits signed. Es una declaracion

# sasbemos que 16 bits van de $-2^(15) hasta 2^(15)-1

len(a)
```

```python
# el mas pequeno 
i=-2**(15) # inferior
a.append(i)

s= 2**15 -1
a.append(s)
print("rango de enteros cortos es",a)
```

```python
# overfloa above error
a.append(2**15)
```

```python
b = 2**15  # 64 bits con signo
print(b)
```

```python
# underflow
a.append(-2**15-1)
```

```python
#Unsigned hal
a = array("H") # unsigned short number

i=0 # inferior
a.append(i)

s = 2**16 -1  # superior
a.append(s)
print("el rango de numeros de precision media (half precision, short) sin signo ess",a)
```

```python
# por encima del mas grande
a.append(2**16)
```

```python
# por debajo del minimo
a.append(-1)
```

```python
# Enteros de 32 bits
a = array('i')

# el rango debe ir desde -2^(31) hastsa 2^(31)-1
i=-2**(31)  # inferior
a.append(i)


# superior
s=2**31 -1 
a.append(s)

print("el rango de numeros de precision completa (single precision) sin signo ess",a)
```
