## Diagramas de Pastel (Pie Charts). 

```python
import matplotlib.pyplot as plt


etiquetas="Uno", "Dos", "Tres", "Seis a Nueve", "Diez y más"
sizes = [28.7, 23.5, 16.6, 7.7, 8.9]
colores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'r']
explotar = (0, 0, 0, 0,  0.4)  # the last one

# plot
plt.title("Cuantos libros leen los colombianos al "
       "año. (Alberto Salcedo, El Tiempo 2016)")

plt.pie(sizes, explode=explotar, labels=etiquetas, colors=colores, 
        autopct='%1.2f%%', shadow=True, startangle = 140)

plt.axis('equal')
plt.show()
```

# Bar Charts (diagramas de barras)

```python
import numpy as np

etiquetas = ('Uno', 'Dos', 'Tres', 'Seis a Nueve', 'Diez y mas')
y_pos = np.arange(len(etiquetas)) # [0, 1, 2, 3, 4], para mi es x-pos posicion en el eje x
# y_pos=[0, 4, 3, 10, 12]
sizes = [28.7, 23.5, 16.6, 7.7, 8.9]


plt.bar(y_pos, sizes, align='center' ,alpha=1.0, width=0.2) #alpha= transaparency 
plt.xticks(y_pos, etiquetas)
plt.ylabel('porcentage')  # realmente no suman cien, o sea que no es porcentaje.
plt.title("Cuantos libros leen los colombianos al "
       "año. (Alberto Salcedo, El Tiempo 2016)")
```

# Diagramas de dispersion (scatter plots)

```python
N=60
g1 = (0.6 + 0.7* np.random.rand(N), np.random.rand(N))
g2 = (0.4 + 0.4* np.random.rand(N), np.random.rand(N))
g3 = (0.4* np.random.rand(N), 0.3*np.random.rand(N))
```

```python
data = (g1, g2, g3)
colors='red', 'green', 'blue'
groups = 'petroleo', 'gas', 'carbon'

for data, color, group in zip(data, colors, groups):
    x, y = data
    plt.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.title("Ejemplo de scatter plot")
plt.legend(loc=2)
```
