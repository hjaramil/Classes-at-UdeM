```python
# Graficas en Python
import matplotlib, numpy
from numpy import pi, sin, cos, arange
from matplotlib.pyplot import plot, xlabel, ylabel, title, axis, grid, show , legend
```

```python
x=arange(0, pi, 0.01)
x
```

```python
plot(x, sin(x), label="$\sin x$")
```

```python
# si quisieras ver esta grafica en una ventana a parte
%matplotlib qt
plot(x, sin(x), label="$\sin x$")
```

```python
import matplotlib.pyplot as plt
plt.plot(x, sin(x))
```
