## Vectorizacion: Evaluamos la formula

$$ S = \sum_{i=0}^{100} \sqrt{ \frac{i \pi}{100}} \cos \frac{i \pi}{100} $$

````python
from numpy import sqrt, cos, pi
# impementacion regular no vectorizada
# x=0
suma=0.0
n=101
for i in range(n):
    x = i*pi/100 
    suma = suma + sqrt(x)*cos(x)
    # x = x + 0.01*pi  # cada vez suma un pi/100

print(suma)
````

````python
# python permite implemntaiones vectorizadas rapidas
x = np.arange(0.0, 1.001*pi, 0.01*pi)
print(sum(sqrt(x)*cos(x)))
````

````python
A = np.array([[1,3], [2,5]])
B = np.array([[-3,4],[2,-1]])


# version de suma de matrices no vectorizada
def sume(A,B):
    n=len(A)
    C = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            C[i,j]=A[i,j]+B[i,j]

    return C

    
````

````python
sume(A,B)
````

### Dot product. Producto de matrices
Si $A,B$ son matrices de forma que el numero de columnas de $A$ es igual al numero de filas de $B$ $n$, se puede hacer el producto $C= AB$

$$ c_{ij} = \sum_{k=1}^n a_{ik} b_{kj}$$

````python
np.dot(A,B)
````

## producto exterior
$A=(1,2)$, $B=(3,4)$

$A \times B = \{ (1,3), (1,4), (2,3), (2,4) \}$

\begin{eqnarray}
A \otimes B =  \begin{pmatrix}
1 \times 3 & 1 \times 4 \\
2 \times 3 & 2 \times 4
\end{pmatrix}
\end{eqnarray}


Matematicamente

Si $A$ es un vector, $B$ es otro vector de la misma dimension
$C=A \otimes B$, si

$c_{ij} = a_i b_j$

## Tensores: 
Tienen 3 interpretaciones distintas:

1.  Geometria diferencial (matematica). Deep learning muestro esto, como parte del algebra multi-lineal.
2.  Fisica (las leyes fisicas se escriben con tensores que muestran fenomenos independientes del sisteam coordenado que se use).
3. Computacional. La que vemos en este curso

3-D Tensors (rank 3 tensors). Rango 3.

Que es el rango?

* Los escalares tienen rango 0
* Los vectores tienen rango 1
* Las matrices tienen rango 2
* Los tensores que muestro a continuacion tienen rango 3.
Con estos ejemplos podemos intuir el significado de rango. Matematicamente
se puede definir como el numero de indices "libres"
$a_{ijkll}$ como $l$ es repetido, no es libre, hay tres indice libres $i,j,k$ este es un tensor de rango 3. Que se hace con el indice "dummy" (repetido)? se suma sobre el. Es decir (notacion de Einstein)

$$a_{ijkll} \rightarrow   \sum_l a_{ijkl} = d_{ijk} $$
