# Revisitamos la clase 11 en los matrix scatter plots

```python
# generemos datos para diagramas de dispersion
import numpy as np
import matplotlib.pyplot as plt
N=60
g1 = (0.6 + 0.7*np.random.rand(N), np.random.rand(N))
g2 = (0.4 + 0.4*np.random.rand(N), np.random.rand(N))
g3 = (0.4*np.random.rand(N), 0.3*np.random.rand(N))

data = (g1, g2, g3)
colors = 'red', 'green', 'blue'
groups =  'petroleo', 'gas', 'carbon'


plt.title("Ejemplo de un scatter plot")
for data, color, group in zip(data, colors, groups):
    x,y = data
    plt.scatter(x,y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.legend(loc=2)
```

### Matrix of scatter plots

```python
import pandas as pd
import seaborn as sns

datos1 = np.transpose([g1[0]])
datos2 = np.transpose([g1[1]])
datos3 = np.transpose([g2[0]])
datos4 = np.transpose([g2[1]])

df = pd.DataFrame(np.hstack((datos1,datos2, datos3, datos4)))

sns.pairplot(df)
```
