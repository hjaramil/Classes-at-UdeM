# Copiar un arreglo

````python
a = np.array([1,2,3])
b=a  # shallow copy (copia superficial)
print(a)
print(b)
print(id(a), id(b))
````

````python
c = a.view()  # la direccion en memoria cambia, aun asi es "shallow"
print(a,c)
print(id(c))
````

````python
a[2] = -10
print(a)
print(b)
print(c)
````

````python
# como hacer para que la copia no sea "shallow", que si cambio el uno no se cambie el otra
a=[1,5,7]
d=a.copy()
print(a,d)
print(id(a), id(d))
````

````python
a[0]=-10
print(a,d)
````
