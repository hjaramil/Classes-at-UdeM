## Diccionarios:
Los diccionarios son objetos (conjuntos) de parejas donde el primer elemento es una "llave" (key) y el segundo el "valor".  Las llaves siempre estan entre comillas, excepto si son numericas. Los valores, si son "strings" entre comillas, si son numeros no comillas, la sintaxis es:

```key : value``` 
para cada elemento del conjunto. Es decir

$$\{ \texttt{key}_1 : \texttt{value}_1, \texttt{key}_2: \texttt{value}_2, \cdots, \texttt{key}_n : \texttt{value}_n \} $$

````python
miDiccionario = {
    "nombre": "Jose",
    "cargo" : "profesor",
    "edad" : "54 años"
}
print(type(miDiccionario))
````

````python
miDiccionario
````

````python
# es mutable un diccionario?, yes
# asi se accesan los elementos del diccionary
miDiccionario['nombre']
````

````python
miDiccionario['nombre'] = "Juan"
miDiccionario['nombre']
````

````python
# concatenemos diccionarios
midic2 = {
    "identificacion":32344554
}

ambos = { **miDiccionario, **midic2}  # concatenacion
ambos
````

````python
# otro = {"color"}
tresDiccionarios = {**ambos, **miDiccionario, **{"color":"rojo"}}  # concatenacion
````

````python
tresDiccionarios
````

````python
diccionario = {}
for i in range(10):
    dict1 = {str(i):i}
    # print(dict1)
    diccionario = {**diccionario, **dict1}


print("diccionario=", diccionario)
len(diccionario)
````

````python
# un diccionario no puede tener claves repetidas:
# ejemplo
diccionario = {}
for i in range(10):
    dict1 = {"numero":i}
    # print(dict1)
    diccionario = {**diccionario, **dict1}


diccionario
````

## Conditional ```if```

Syntaxis:
\begin{eqnarray}
\texttt{if  statement:} \\
\texttt{do this} \\
\texttt{keep working}
\end{eqnarray}

````python
a=2.0
b=3.5
if a<b:
    print("a es menor que b")

print("seguimos por aca")

````

## Conditional ```if else```.

Syntaxis:
\begin{eqnarray}
&&\texttt{if  statement:} \\
   && \quad \texttt{do this} \\
&&\texttt{else:} \\
&& \quad \texttt{do that} \\
&&\texttt{keep working}
\end{eqnarray}

````python
a=2.0
b=2.0

if a > b:
    print(" a es mayor que b")
else:
    print("a es menor o igual a b")
````

## welcome to the ```range()``` method
```range()``` es un tipo nuevo. Es una lista de enteros, con o sin saltos abierta a derecha en los enteros. Pude ser ascendente o descendente (paso <0)

La sintaxis:

```range(in,fin,paso)```.
Aca  
* ```in``` es el valor inicial
* ```fin``` es el valor final (no includo, por que es abierto a derecha
* ``` paso ``` es el incremento , decremento si es negativo

cuando ```in``` no esta, se asume 0, 

````python
a= range(10) # range es built-in
print(a)
````

````python
print(type(a))
````

````python
list(range(-1,5,2))
````

````python
list(range(5, -10, -3))
````
