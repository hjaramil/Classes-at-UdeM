# Multivariate Random Variables: Variables aleatortias multivariadas

Sea $X_1, X_2, \cdots, X_n$ un conjunto de variables aleatorias discretas. Podemos asociar una densidad de probabilidad multivariada que se llama
**probabilidad conjunta** (join density). Los mismos principios axiomaticos de la probabilidad univariada se aplican.
Veamos:

Primero seamos claros con la notacion
$$ p(x_1, x_2, \cdots, x_n) = P_{X_1 X_2 \cdots X_n}(X_1 = x_1, X_2 = x_2, \cdots, X_n = x_n)$$.

Los axiomas, que se extienden.

* $p(x_1, x_2, \cdots, x_n) \ge 0 \quad , \quad ,  \forall x_1, x_2, \cdots , x_n $

* $\sum_{x_1} \sum_{x_2} \cdots \sum_{x_n} p(x_1, x_2 \cdots, x_n) = 1 $
de otra forma $P(S)=1$ donde $S$ es el universo.

* $P([X_1, X_2, \cdots , X_n]) \in A] = \sum_{(x_1, x_2, \cdots, x_n) \in A} p(x_1, x_2, \cdots, x_n) $, se asume que los eventos son excluyentes. 
Donde se asume $A \subset \mathbb{R}^n$, que a la vez es un subconjunto del universo $S$.

Si $X_1, X_2, \cdots, X_n$ son variables continuas:

* $p(x_1, x_2, \cdots, x_n) \ge 0 \quad , \quad ,  \forall x_1, x_2, \cdots , x_n $
* $\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \cdots \int_{-\infty}^{\infty} p(x_1, x_2 \cdots, x_n) dx_1 dx_2 \cdots d x_n = 1 $
de otra forma $P(S)=1$ donde $S$ es el universo.

En particular si $n=2$ hablamos de una distribucion **bivariate** (bivariada).
$p(x,y) = P(X=x, Y=y) = P( X=x \cap Y=y)$.

Introducimos la **conditional join density**. La probabilidad conjunta condicionada. 

$$ p(y | x) =  \frac{p(x,y)}{p(x)}$$

de aca que

$$ p(x,y) = p(y|x) p(x) $$

De la misma forma que derivamos (probamos) la regla de la multiplicacion  (la discutimos en le problema de induccion de la tarea, al comienzo de esta clase)

$$ p(x_1, x_2, \cdots, x_n) = \prod_{t=2}^n p(x_t | x_{t-1}, x_{t-2}, \cdots, x_1) p(x_1)$$

Si los eventos $x_t, t=2,3, \cdots, n$ sosn independientes de los previos $x_t, t=1,2, \cdots, n-1$ entonces $p(x_t | x_{t-1}, x_{t-2}, \cdots, x_1) = p(x_t)$ de forma que

$$ p(x_1, x_2, \cdots, x_n) = \prod_{t=1}^n p(x_t). $$

Mas adelante, vamos a hablar de **likelihood** (verosimilitud). 
Aca se parametrizan las densidades de acuerdo a un parametro $\theta$ y escribimos

$$ p(x_1, x_2, \cdots, x_n | \theta) =  \prod_{t=2}^n p(x_t | x_{t-1}, x_{t-2}, \cdots, x_1; \theta) p(x_1 | \theta)  \quad  (*) $$

**Ejemplo**: Estudiamoso la densidad de probabilidad multi-variada normal.

$$ p(x, \mu, \Sigma) = \mathcal{N}(\mu, \Sigma) = 
\frac{1}{\sqrt{(2 \pi)^n | \det(\Sigma) |}}  \mathrm{e}^{-\frac12 (x-\mu)^T \Sigma^{-1} (x - \mu)}$$

$x \in \mathbb{R}^n$, $\mu \in \mathbb{R}^n$. Quien es $\Sigma$?
A $\Sigma$ se le llama la matriz de **covarianza**. 

Por ejemplo, asumamos que $X \in \mathbb{R}^{m \times n}$, como  en el curso de ML, tenemos $n$ columnas (datos) cada una con $m$ dimensiones. 
Que los vectores $X_1, X_2, \cdots, X_n$ conforman la matriz $X$. Con esta matriz $X construimos la matriz de covarianza

$$ \Sigma = \frac1n (X-\mu)^T   (X -\mu ) $$

Discusion:  $ X \in \mathbb{R}^{m \times n}$ segun esto  $\mu$ tambien deberia estar en ese espacio? como esto? $\mu$ es una matriz $m \times n$ donde
cada columna $i$ tiene la media del vector  $X_i$.
A la matriz $\Sigma^{-1}$ se le llama ** matriz de concentracion ** .

Como un pie de pagina:

$$ \Sigma_{ij} = \mathrm{cov}[ X_i , X_j] = \mathbb{E}[ (X_i-\mu_i)^T (X_j - \mu_j) $$

El caso mas simple de matriz de covarianza es el de una diagonal.

\begin{eqnarray*}
  \Sigma= \begin{pmatrix}
    \sigma_1^2 & 0 & \cdots & \cdots & \cdots & 0  \\
    0  & \sigma_2^2 & \ddots &  \cdots & \cdots & \vdots  \\
    \vdots  & \ddots & \ddots &  \ddots & \ddots & \vdots \\
    \vdots  & \ddots & \ddots &  \ddots & \ddots & \vdots \\
    \vdots  & \ddots & \ddots &  \ddots & \sigma_{n-1}^2 & 0 \\
    0 & \cdots & \cdots & \cdots & 0 & \sigma_n^2
  \end{pmatrix}.
\end{eqnarray*}

Las dimensiones $X$ es $m \times n$, entonces $(X-\mu)^T (X-\mu)$ es $n \times n$. La matriz de covarianza es cuadrada. 

La inversa es:


\begin{eqnarray*}
  \Sigma^{-1}=  \begin{pmatrix}
    \sigma_1^{-2} & 0 & \cdots & \cdots & \cdots & 0  \\
    0  & \sigma_2^{-2} & 0 &  \cdots & \cdots & \vdots  \\
    \vdots  & \ddots & \ddots &  \ddots & \ddots & \vdots \\
    \vdots  & \ddots & \ddots &  \ddots & \ddots & \vdots \\
    \vdots  & \ddots & \ddots &  \ddots & \sigma_{n-1}^{-2} & 0 \\
    0 & \cdots & \cdots & \cdots & 0 & \sigma_n^{-2}
  \end{pmatrix}.
\end{eqnarray*} 

Que la matriz de covarianza sea diagonal implica que los vectores $X_i$ son independientes (estadisticamente). Es decir, no hay **correlacion** entre
$X_i$ y $X_j$ para $i \ne j$.  Si no hay correlacion, es facil ver
que la probabilidad normal multivariada viene de multiplicar las $n$ probabilidades univariadas.

Veamos esto.

$$ p(x, \mu, \Sigma) = \mathcal{N}(\mu, \Sigma) = 
\frac{1}{\sqrt{(2 \pi)^n | \det(\Sigma) |}}  \mathrm{e}^{-\frac12 (x-\mu)^T \Sigma^{-1} (x - \mu)}$$

Comencemos con el determinante.

$$ \det \Sigma = \prod_{j=1}^n \sigma_j^2 $$
de forma que el coeficiente de la ecuacion de arriba


$$ \frac{1}{\sqrt{(2 \pi)^n | \det \Sigma |}} =  \frac{1}{\sqrt{(2 \pi)^n | \det \Sigma|}}  $$
$$ \frac{1}{\sqrt{(2 \pi)^n | \det \Sigma |}} =  \frac{1}{\sqrt{(2 \pi)^n | \prod_{j=1}^n \sigma_j^2 |}}  $$

Vamos ahora al exponente

$$ -\frac12 (x-\mu)^T \Sigma^{-1} (x - \mu)= -\frac12 \sum_{j=1}^n \frac{(x_i - \mu_i)^2}{\sigma_j^2}$$

De otro lado, si multiplicamos todas la probabilidades univariadas encontramos:

\begin{eqnarray} p(x, \mu, \Sigma) &=& p(x_1) p(x_2) \cdots p(x_n) \\
\frac{1}{ \sqrt{(2 \pi)^n \prod_{j=1}^n \sigma_j^2} }
\mathrm{e}^{-\frac12 \sum_{j=1}^n \frac{(x_j -\mu_j)^2}{\sigma_j^2} }
\end{eqnarray}

De forma que la probabilidad multivariadad es la productoria de las probabilidades univariadas. 

## Variables aleatorias multivariadas de acumulacion. Cumulative Distributive Functions (CDF)s.

Sean $X_1, X_2, \cdots, X_n$ variables aleatorias


\begin{eqnarray} F_{X_1 X_2 \cdots X_n}(x_1, x_2, \cdots, x_n ) &=& P(X_1 \le x_1, X_2 \le x_2, \cdots, X_n \le x_n  \\
&=& P(X_1 \le x_1 \text{ and } X_2 \le x_2 \text{ and } \cdots \text{ and }  X_n \le x_n  \\
&=& P[ \cap_{i=1}^n ( X_i \le x_i])
\end{eqnarray}

Otro nombre de esto es **join cumulative distributive function** (la funcion distributiva acumulada conjunta).

De aca podemos obtener probabilidades **marginales**. Por ejemplo, la marginal en $x_1$ es asumiendo que todas las demas siempre son ciertas.

$$ F_{X_1}(x_1) =P(X_1 \le x_1, X_2 \le \infty, \cdots, X_n \le \infty) = P(X_1 \le x_1) $$

Otro ejemplo. Marginales en $x_1, x_2$.

$$ F_{X_1}(x_1, x_2) =P(X_1 \le x_1, X_2 \le x_2, \cdots, X_n \le \infty) = P(X_1 \le x_1, X_2 \le x_2) $$

y en general escojamos $X_i$, $i \in I$. 

$$F_{X_I}(x_I)= P[ \cap  ( X_i \le x_i)]  $$
