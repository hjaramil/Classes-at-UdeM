# Modelo AR(2), AR(p), p=2
Este modelo se basa en la ecuacion

$$X_t = c + \phi_1 X_{t-1} + \phi_2 X_{t-2} + Z_t \quad (1) $$

Es necesario ejercitar el algebra para esta seccion. Especialmente recuerde


$$ \frac{1}{1-x} = 1 + x + x^2 + \cdots x^k + ... \quad (0) $$
y $|x|<1$.

En la ecuacion (1) usamos el backshift operator

$$ X_t = c + (\phi_1 B + \phi_2 B^2) X_t + Z_t$$
agrupemos

$$ (1 - \phi_1 B - \phi_2 B^2) X_t = c + Z_t  \quad (2) $$.
donde el termino de la derecha es el "forcing term" .
Esta es una ecuacion en diferencias (la incognita es $X_t$, modelamiento).
Recuerden la teoria de ecuaciones en diferencias, 

* Hallar una solucion particular
* Hallar una solucion al sistema homogeno (forcing term is 0)
* La combinacion de las dos, con las condiciones iniciales resuelve el problema.

Extendemos el analisis de AR(1) a AR(2) donde "despejamos" $X_t$ de la Ecuacion (2). 

$$ X_{tp} = \frac{1}{1 - \phi_1 B - \phi_2 B^2} (c + Z_t) =
(1 - \phi_1 B - \phi_2 B^2)^{-1} (c + Z_t)  \quad (-1) $$
Cuando existe el inverso? vamos a investigar esto en esta clase.

De nuevo, trayendo la teoria de ecuaciones en diferencias, para solucionar este problema debemos hallar las raices del polinomio caracteristico

$$ p(B) = 1 - \phi_1 B - \phi_2 B^2 \quad, \quad (2A0) $$
Debemos hallar las raices $r_1, r_2$. Pueden ser distintas o iguales (repetidas), reales o complejas.

Factorizamos el polinomio caracteristico

$$1 - \phi_1 B - \phi_2 B^2 = (1 - \varphi_1 B)(1 - \varphi_2 B) $$ .

Las raices del polinomio en $B$ serian

## Las raices sean distintas, $r_1 \ne r_2$.

$$ r_1 = \frac{1}{\varphi_1}  \quad r_2 = \frac{1}{\varphi_2} $$

De forma que podemos escribir

$$p(B) = (1 - B/r_1) (1 - B/r_2)  $$
Nosotros sabemos invertir $1-B/r_1$, $1-B/r_2$ (basados en la ecuacion 0) pero como invertimos el producto? Usamos fracciones parciales.

El metodo de **fracciones parciales** para un polinomio de orden 2 factorizado dicta que


$$\frac{1}{1 - \phi_1 B - \phi_2 B^2} = \frac{A_1}{1 - B/r_1} + \frac{A_2}{1 - B/r_2}  \quad (2A)$$

$A_1, A_2$ se deben determinar

$$ \frac{1}{1 - \phi_1 B - \phi_2 B^2} = \frac{A_1(1 - B/r_2) + A_2(1 - B/r_1}{1 - \phi_1 B - \phi_2 B^2}$$
Como los denominadores son iguales, los numeradores tambien lo son. Hacemos un
apareo (matching).  

$$ 1 = A_1(1 - B/r_2) + A_2(1 - B/r_1) $$  (polinomios en $B$).


\begin{eqnarray}
1 &=& A_1 + A_2 \quad \text{coeficiente de orden 0} 
\quad , 
\quad (3) \\
0 &=& -\frac{A_1}{r_2} - \frac{A_2}{r_1} . 
\end{eqnarray}

Toca resolver este sistema de 2 equaciones con 2 incoginitas.
De la primera ecuacion $A_1 = 1 - A_2$.  Insertamos esto en la segunda
y tenemos

$$ 0 = -\frac{1-A_2}{r_2} - \frac{A_2}{r_1} = -
\frac{1}{r_2} + \frac{A_2}{r_2} - \frac{A_2}{r_1} = -\frac{1}{r_2} + A_2 \left (  \frac{1}{r_2} - \frac{1}{r_1} \right )$$

Despejamos de $A_2$,

$$A_2 = \frac{\frac{1}{r_2}}{(1/r_2 - 1/r_1)} =\frac{1}{r_2 \frac{r_1-r_2}{r_2 r_1}} = \frac{r_1}{r_1 - r_2} $$

De la ecuacion (3)

$$A_1 = 1 - A_2 =1 - \frac{r_1}{r_1 - r_2} = \frac{r_1 - r_2 - r_1}{r_1 - r_2} = - \frac{r_2}{r_1 - r_2}  $$
Con esto ya encontramos las fracciones parciales.

De forma que

\begin{eqnarray}
\frac{1}{1 - \phi_1 B - \phi_2 B} &=& -\frac{r_2}{(r_1 - r_2)(1 - B/r_1)} + \frac{r_1}{(r_1-r_2)(1 - B/r_2)}  \\
&=& \frac{1}{r_1 - r_2}  \left [  \frac{r_1}{1 - B/r_2} - \frac{r_2}{1 - B/r_1}  \right ] \\
&=& \frac{1}{r_1 - r_2}  \left [  r_1 \left ( 1  + (B/r_2) +(B/r_2)^2 + \cdots + (B/r_2)^n + \cdots  \right ) -r_2 \left ( 1 + (B/r_1) + (B/r_1)^2 + \cdots + (B/r_1)^n + \cdots \right ) \right ] \\
&=&\frac{1}{r_1 - r_2}   \left [  (r_1 - r_2) + \left ( r_1 \frac{1}{r_2} - r_2 \frac{1}{r_1} \right ) B  + \left ( r_1 \frac{1}{r_2^2} - r_2 \frac{1}{r_1^2} \right ) B^2 + \cdots + \left ( r_1 \frac{1}{r_2^n} - r_2 \frac{1}{r_1^n} \right ) B^n + \cdots \right  ]
\end{eqnarray}



Se asume que $|r_1|>1$, $|r_2|>1$. Dejamos de tarea explicar que pasa si $|r_1|=1$ o $|r_2|=1$.

En el apendice A de las notas probamos que las condiciones para convergencia (sea estacionaria) son

$$ \phi_1 + \phi_2 < 1  \quad , \quad \phi_2 - \phi_1 < 1 \quad, \quad |\phi_2| < 1 $$. La region de convergencia tiene una grafica que se llama
el **Triangulode Stralkowski**. Lo mostramos enseguida.

$$X_t = [ 1 + B/r + B^2/r^2 + \cdots + (B/r)^n + \cdots ] Y_t$$

En resumen
$X_t = Op^2 Z_t = Op (Op Z_t) $

De forma que tenemos

$$Y_t = Z_t + Z_{t-1}/r + \cdots + Z_{t-n}/r^n + \cdots  $$
