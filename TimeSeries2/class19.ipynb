# Yule-Walker Equations:
Tienen como objeto encontrar los coeficientes $\phi_i$, $i=1,2, \cdots,$p

El modelo AR(p) es

$$ X_t = \sum_{i=1}^{\text{p}} \phi_i X_{t-i} + Z_t.$$
Asumimos $c=0$.

Por pasos:

* Lag 1: Calcula


\begin{eqnarray}
X_t X_{t+1} &=& X_t \left (  \sum_{i=1}^{\text{p}} \phi_i X_{t-i+1} + Z_{t+1 }\right ) \\
&=& \sum_{i=1}^{\text{p}} \phi_i X_t X_{t-i+1} + X_t Z_{t+1}
\end{eqnarray}
Tomando la esperanza encontramos $\gamma_1$.

$$\gamma_1 = \sum_{i=1}^{\text{p}} \phi_i \mathbb{E}[X_t X_{t-i+1}] + \mathbb{E}[X_t Z_{t+1}] =  \sum_{i=1}^{\text{p}} \phi_i \mathbb{E}[X_t X_{t-i+1}] = \sum_{i=1}^{\text{p}} \phi_i \gamma_{i-1} $$
Dividimos todo por $\gamma_0 = \sigma^2$ y encontramos

$$ \rho_1 = \sum_{i=1}^{\text{p}} \phi_i \rho_{i-1}  $$

* Lag 2: Calcula


\begin{eqnarray}
X_t X_{t+2} &=& X_t \left (  \sum_{i=1}^{\text{p}} \phi_i X_{t-i+2} + Z_{t+2 }\right ) \\
&=& \sum_{i=1}^{\text{p}} \phi_i X_t X_{t-i+2} + X_t Z_{t+2}
\end{eqnarray}
Tomando la esperanza encontramos $\gamma_2$.

$$\gamma_2 = \sum_{i=1}^{\text{p}} \phi_i \mathbb{E}[X_t X_{t-i+2}] + \mathbb{E}[X_t Z_{t+2}] =  \sum_{i=1}^{\text{p}} \phi_i \mathbb{E}[X_t X_{t-i+2}] = \sum_{i=1}^{\text{p}} \phi_i \gamma_{i-2} $$
Dividimos todo por $\gamma_0 = \sigma^2$ y encontramos

$$ \rho_2 = \sum_{i=1}^{\text{p}} \phi_i \rho_{i-2}  $$

* Lag $k$, $k=1,2,, \cdots \text{p}$: Calcula


\begin{eqnarray}
X_t X_{t+k} &=& X_t \left (  \sum_{i=1}^{\text{p}} \phi_i X_{t-i+k} + Z_{t+k }\right ) \\
&=& \sum_{i=1}^{\text{p}} \phi_i X_t X_{t-i+k} + X_t Z_{t+k}
\end{eqnarray}
Tomando la esperanza encontramos $\gamma_1$.

$$\gamma_1 = \sum_{i=1}^{\text{p}} \phi_i \mathbb{E}[X_t X_{t-i+k}] + \mathbb{E}[X_t Z_{t+k}] =  \sum_{i=1}^{\text{p}} \phi_i \mathbb{E}[X_t X_{t-i+k}] = \sum_{i=1}^{\text{p}} \phi_i \gamma_{i-k} $$
Dividimos todo por $\gamma_0 = \sigma^2$ y encontramos

$$ \rho_k = \sum_{i=1}^{\text{p}} \phi_i \rho_{i-k} . $$
Esto es valido para $k=1,2, \cdots, \text{p}$.

Esta ecuacion general la podemos escribir de forma matricial. 

\begin{eqnarray}
\begin{pmatrix}
\rho_0 & \rho_1 & \rho_2 & \cdots & \rho_{\text{p}-1} \\
\rho_1 & \rho_0 & \rho_1 & \cdots & \rho_{\text{p}-2} \\
\vdots & & & \cdots  & \vdots \\
\vdots & & & \cdots  & \vdots \\
\rho_{\text{p}-1}& \rho_{p-2} & \rho_{p-3} & \cdots & \rho_{0} \\
\end{pmatrix}
\begin{pmatrix}
\hat{\phi_1} \\
\hat{\phi_2} \\
\vdots \\
\vdots \\
\hat{\phi_{\text{p}}}  
\end{pmatrix}
= \begin{pmatrix} 
\rho_1  \\
\rho_2 \\
\vdots \\
\vdots \\
\rho_p
\end{pmatrix} 
\end{eqnarray}

Este sistema se puede resolver para los $\hat{\phi_i}$ conociendo los $\rho_i$, $i=1,2, \cdots, \text{p}$. 

Recuerden que $\rho_0=1$ de forma que podemos escribir este sistema como:

\begin{eqnarray}
\begin{pmatrix}
1 & \rho_1 & \rho_2 & \cdots & \rho_{\text{p}-1} \\
\rho_1 & 1 & \rho_1 & \cdots & \rho_{\text{p}-2} \\
\vdots & & & \cdots  & \vdots \\
\vdots & & & \cdots  & \vdots \\
\rho_{\text{p}-1}& \rho_{p-2} & \rho_{p-3} & \cdots & 1 \\
\end{pmatrix}
\begin{pmatrix}
\hat{\phi_1} \\
\hat{\phi_2} \\
\vdots \\
\vdots \\
\hat{\phi_{\text{p}}}  
\end{pmatrix}
= \begin{pmatrix} 
\rho_1  \\
\rho_2 \\
\vdots \\
\vdots \\
\rho_p
\end{pmatrix} 
\end{eqnarray}

Lo que tenemos es un sistema 

$$ R \hat{\phi} = r $$
a resolver para $\hat{\phi}$ donde:



\begin{eqnarray}
R = \begin{pmatrix}
1 & \rho_1 & \rho_2 & \cdots & \rho_{\text{p}-1} \\
\rho_1 & 1 & \rho_1 & \cdots & \rho_{\text{p}-2} \\
\vdots & & & \cdots  & \vdots \\
\vdots & & & \cdots  & \vdots \\
\rho_{\text{p}-1}& \rho_{p-2} & \rho_{p-3} & \cdots & 1 \\
\end{pmatrix} \quad , \quad
\hat{\phi} = \begin{pmatrix}
\hat{\phi_1} \\
\hat{\phi_2} \\
\vdots \\
\vdots \\
\hat{\phi_{\text{p}}}  
\end{pmatrix} \quad , \quad 
r = \begin{pmatrix} 
\rho_1  \\
\rho_2 \\
\vdots \\
\vdots \\
\rho_p
\end{pmatrix}
\end{eqnarray}

De nuevo, a este sistema se le conoce como las Yule-Walker equations.



A las matrices donde una fila circula para producir las demas, tal y como se ve en la matriz $R$ se les llama 

[Toeplitz matrix](https://es.wikipedia.org/wiki/Matriz_de_Toeplitz).

La solucion del sistema es

$$\hat{\phi} = R^{-1} r $$
pues $R$ tiene inversa por que es positiva definida (esto no lo probamos aca).

Lo interesante es que este problema se encuentra en otras areas de la ciencia con otro nombre. Por ejemplo en el procesamiento de datos digitales (Digital Signal Processing) existe un metodo que se llama

[Wiener Filtering](https://en.wikipedia.org/wiki/Wiener_filter).

La idea es la siguiente:
Para la nosotros en el texto tenemos la ecuacion (4.1)

$$y_i = \sum_{k=1}^{\text{p}} x_{i-k} h_k = (x \ast h) $$
En el caso particular del AR(p)$ tenemos que

$$ y_i =  x_{i} \approx  \sum_{k=1}^{\text{p}} x_{i-k} h_k  $$

Una forma de encontrar los coeficientes $h_k$ (que para nosotros son $\phi_k$ es
con **minimos cuadrados**.

Definimos un error

$$\epsilon_n = y_n - x_n $$
donde $y_n$ es el momdelo de arriba y $x_n$ son los datos que ya tenemos.
El problema de minimos cuadrados consistge en encontrar la varianza de este error y la minizamos. A esto se conoce como el

[Mean Square Error (MSE)](https://en.wikipedia.org/wiki/Mean_squared_error).

Se deja como tarea usar este metodo descrito aca, del MSE para encontrar
la ecuacion de Yule-Walker arriba. 

A continuacion mostramos un ejemplo de como hallar los coeficientes 
$\hat{\phi_i}$ a partir de la Yule-Walker Equation.
El ejemplo es de orden de 3, es decir AR(3):

$$ X_t = \phi_1 X_{t-1} + \phi_2 X_{t-2} + \phi_3 X_{t-3} + Z_t $$,
$c=0$.



```python
# ejemplo AR(3)
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# parametros <1 en valor absoluto
phi1 = 0.8
phi2 = -0.3
phi3 = 0.1

# definamos el ruido blanco
np.random.seed(1)
mu, sigma = 0, 1
N=1000
Z=np.random.normal(mu, sigma, N)

# inicializacion de la serie de tiempo
X = np.zeros(N)

# condiciones iniciales. 3
X[0] = Z[0]
X[1] = Z[1]
X[2] = Z[2]

# la generacion de datos recursiva
for t in range(3,N):
    X[t] = Z[t] + phi1*X[t-1] + phi2*X[t-2] + phi3*X[t-3]
                         

# graficas
plt.xlabel("numero de muestra")
plt.ylabel("valor de la muestra")
plt.title(f"proceso AR(3) con $\phi_1$={phi1}, $\phi_2$={phi2}, $\phi_3$={phi3}")
plt.plot(X)
plt.show();



```

### Construccion del sistema Yule-Walker
\begin{eqnarray}
\begin{pmatrix}
1 & \rho_1 & \rho_2 & \cdots & \rho_{\text{p}-1} \\
\rho_1 & 1 & \rho_1 & \cdots & \rho_{\text{p}-2} \\
\vdots & & & \cdots  & \vdots \\
\vdots & & & \cdots  & \vdots \\
\rho_{\text{p}-1}& \rho_{p-2} & \rho_{p-3} & \cdots & 1 \\
\end{pmatrix}
\begin{pmatrix}
\hat{\phi_1} \\
\hat{\phi_2} \\
\vdots \\
\vdots \\
\hat{\phi_{\text{p}}}  
\end{pmatrix}
= \begin{pmatrix} 
\rho_1  \\
\rho_2 \\
\vdots \\
\vdots \\
\rho_p
\end{pmatrix} 
\quad, \quad (1)
\end{eqnarray}



Lista de pasos:

* De los datos calculamos los $\rho_i$. Es decir necesitamos la ACF. Esta es a penas una de las razones por las cuales las ACF es importante. Otro, por ejemplo, es que para los modelos MA (Moving Average) que veremmos a partir del proxima semana, el ACF me determina el orden "q" del MA. 

* constuir la matriz $R$ y el vector $r$. 

* invertir $R$, hallar $R^{-1}$
* hallar $\hat{\phi}=R^{-1} r$.

```python
# hallemos la ACF de los datos $X_t$.
M=3 
Rho = np.zeros((M,M))

# el vector rho se calcula con statsmodels.
rho = sm.tsa.stattools.acf(X)[0:M+1]

# construyamos la matriz Toepliz Rho
for i in range(M):
    for j in range(i, M):
        Rho[i,j]=rho[j-i]  # de la diagnoal en adelante, triangular superior
        if i != j: Rho[j,i]=Rho[i,j]  # el espejo de abajo

rhovec = rho[1:M+1]

# resolvemos el sistema (1)
phivec = np.linalg.solve(Rho, rhovec)
print(phivec)
print(f"$\phi_1$={phi1}, $\phi_2$={phi2}, $\phi_3$={phi3} ")

# no suena muy prometedor. Pero veamos como se calcula usando statmodels
```

```python
rhoYK , sigma = sm.regression.yule_walker(X, order=3, method='mle')
rhoYK
```

Con esto terminamos una parte importante del curso que son los modelos AR(p). Habiamos dicho que el ACF es importante por varias razones:

* Estudia la "auto-similitud" bajo un desplazamiento (shift). Por ejemplo una funcion seno es autosimilar con desplazamientos de $\phi$ y esto se detecta facilmente con la ACF, que deberia mostrar "spikes" cada periodo $\phi$.

* Se usa para calcular los coeficientes del modelo $\phi_i$, $i=1,2, \cdots, p$ asumiendo que este es AR(p).

* Se usa para calcular el orden q de un MA(q). Proximas clases (Moving Average)

**PACF**:  Partial AutoCorrelation Function. Esta es importante para calcular

# Regresion Lineal
## Datos univariados  (una dimension 1D):

Asumamos que tenemos una serie de puntos $(x_1, y_1), (x_2, y_2), \cdots , (x_m, y_m)$ en $\mathbb{R}^2$ los cuales queremos ajustar a una linea
$y = \theta_0 + \theta_1 x$ donde $\theta_0$ es el intercepto y $\theta_1$ es la pendiente. 

### Ecuaciones normales:
Las ecuaciones que establecen esta hipotesis son

\begin{eqnarray}
y_1 &=& \theta_0 + \theta_1 x_1 \\
y_2 &=& \theta_0 + \theta_1 x_2 \\
&\vdots& \\
y_m &=& \theta_0 + \theta_1 x_m
\end{eqnarray}
Esto lo podemos llevar a forma matricial
es decir $y=x \theta$ donde

\begin{eqnarray}
y = \begin{pmatrix}
y_1 \\
y_2 \\
\vdots \\
y_m
\end{pmatrix} \quad , \quad
x = \begin{pmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots & \vdots \\
1 & x_m
\end{pmatrix}
\quad , \quad
\theta = \begin{pmatrix}
\theta_0 \\
\theta_1 
\end{pmatrix}
\end{eqnarray}

Lo vimos en Machine Learning que los minimos cuadrados se resuelven con las ecuaciones normales.

\begin{eqnarray}
x \theta &\approx& y \\
x^T x \theta &\approx& x^T y  \quad \text{estas son las ecuaciones normales} \quad (2) 
\end{eqnarray}

En el caso de regresion $x^T x$ tiene inversa (sino, toca halla la pseudo-inversa o usar Tikhonov regularization, ver ML).
Si $(x^T x)^{-1}$ existe entonces de la ecuacion (2) multiplicando
a ambos lados por $(x^T x)^{-1}$ encontramos que:

$$ \theta = (x^T x)^{-1} x^T y$$. Esta solucion minima el MSE:


$$\min_{\theta} \mathbb{V}[\epsilon] = \min_{\theta} \mathbb{E} [ 
     (X \theta - y)^T (X \theta - y) ] $$

De otra forma simple

$$\min_{\theta} \| X \theta - y \|^2 $$

A esto se le llama regresion lineal.

Para mi esta solucion es elegante y "simple", pero es comun en los textos encontrar la siguiente solucion:



\begin{eqnarray}
\text{slope} &=& \theta_1 = \frac{\sum_{i=1}^m (x_i - \overline{x})(y_i - \overline{y})}{\sum_{i=1}^m (x_i - \overline{x}) (x_i - \overline{x})}
= \frac{\text{Cov}(x,y)}{\mathbb{V}[x]} \\
\text{intercept} &=& \theta_0 = \overline{y} - \theta_1 \overline{x}  \quad (3)
\end{eqnarray}


Donde

\begin{eqnarray}
x &=& \begin{pmatrix}
x_1 \\
x_2 \\
\vdots \\
x_m
\end{pmatrix} \quad , \quad 
y = \begin{pmatrix}
y_1 \\
y_2 \\
\vdots \\
y_m
\end{pmatrix} \\
\overline{x} &=& \frac1m \sum_{i=1}^m x_i \quad , \quad
\overline{y} = \frac1m \sum_{i=1}^m y_i \\
\mathbb{V}[x] &=& \frac{1}{m-1} \sum_{i=1}^m (x_i - \overline{x})^2 \quad, \quad
\text{Cov}(x,y) = \frac{1}{m-1}  \sum_{i=1}^m (x_i - \overline{x}) (y_i - \overline{y})
\end{eqnarray}

De la ecuacion (3) es intersante observar que la linea de regresion **pasa* por el punto $(\overline{x}, \overline{y})$. Es decir la linear de regresion  para por el centro geometrico de los puntos $(x_i)$ y $(y_i)$. 

Veamos esto en python: creamos datos aleatorios y con ellos unos datos AR(1) y luego hacemos la regresion lineal sobre ellos.



```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
mu, sigma = 0, 1
N=1000
Z = np.random.normal(mu, sigma, N)


phi=0.3
X=np.zeros(N)
X[0] = Z[0]

for t in range(1,N):
    X[t] = Z[t] + phi*X[t-1] 

# graficas
plt.xlabel("numero de muestra")
plt.ylabel("valor de la muestra")
plt.title(f"proceso AR(1) con $\phi_1$={phi}")
plt.plot(X)
plt.show();
```

```python
# hacemos un "cross-plot" de graficar X_t versus X_{t-1}
Xt = X[:-1]  # saco la ultima muestra para que el numero de datos sea N-1
Xtm1 = X[1:]  # desde la seguna muestra hasta la ultima, #datos N-1

# graficas
plt.xlabel("numero de muestra")
plt.ylabel("valor de la muestra")
plt.title(f"proceso AR(1) con $\phi_1$={phi}, $X_{t-1}$ versus $X_t$")
plt.scatter(Xtm1, Xt, s=2)
plt.show();
```
