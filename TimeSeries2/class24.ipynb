## PACF de un  modelo MA(q).

\begin{eqnarray*}
  P_k  = 
  \frac{ \det
    \begin{pmatrix}
      1 & \rho_1 & \rho_2 & \cdots & \rho_{{k}-2} & \rho_{1}  \\
      \rho_1 & 1 & \rho_1 & \cdots & \rho_{{k}-3} & \rho_{2}  \\
      \vdots &  & & \cdots &  & \vdots \\
      \vdots &  &  & \cdots &  & \vdots \\
      \rho_{{k}-2} & \rho_{{k}-3} & \rho_{{k}-4} & \cdots & 
      1 & \rho_{k-1} \\
      \rho_{{k}-1} & \rho_{{k}-2} & \rho_{{k}-3} & \cdots & 
      \rho_1 & \rho_k \\
    \end{pmatrix}
  }
  {
    \det 
    \begin{pmatrix}
      1 & \rho_1 & \rho_2 & \cdots & \rho_{{k}-2} & \rho_{{k}-1}  \\
      \rho_1 & 1 & \rho_1 & \cdots & \rho_{{k}-3} & \rho_{{k}-2}  \\
      \vdots &  &  & \cdots &  & \vdots \\
      \vdots &  &  & \cdots &  & \vdots \\
      \rho_{{k}-2} & \rho_{{k}-3} & \rho_{{k}-4} & \cdots & 
      1 & \rho_1 \\
      \rho_{{k}-1} & \rho_{{k}-2} & \rho_{{k}-3} & \cdots & 
      \rho_1 & 1 \\
    \end{pmatrix} 
  } \, .
\end{eqnarray*}


Sabemos que $P_0=\rho_0=1$, $P_1 = \rho_1$ (los mismos del ACF). La dificultad aparece para $k>1$.
Debemos hallar este coeciente de determinantes. Sabmoes tambien que $\rho_k=0$, $k \ge 2$ en el caso de un MA(1).
Estudiemos $P_k$ , $k >1$ y un modelo MA(1).

Como $\rho_k=0$, $k \ge 2$ las matrices tienen muchos ceros.

\begin{eqnarray}
  P_k  = 
  \frac{ \det
    \begin{pmatrix}
      1 & \rho_1 & 0 & \cdots & \cdots & \cdots & 0 & \rho_{1}  \\
      \rho_1 & 1 & \rho_1 & \ddots &  &  & \vdots & 0  \\
      0  & \rho_1 & 1 &  \rho_1 & \ddots &  & \vdots & \vdots  \\
      \vdots & \ddots & \ddots & \ddots & \ddots & \ddots & \vdots & \vdots  \\
      \vdots &  &  \ddots & \ddots & \ddots & \ddots & 0 & \vdots \\
      \vdots &  &  &  \ddots & \rho_1 & 1 & \rho_1 & \vdots  \\
      \vdots & & & & \ddots & \rho_1 & 1 & \vdots  \\
      0 & \cdots & \cdots & \cdots & \cdots & 0  &  \rho_1  & 0 
    \end{pmatrix}
  }
  {
    \det 
    \begin{pmatrix}
      1 & \rho_1 & 0 & \cdots & \cdots & \cdots & \cdots & 0 \\
      \rho_1 & 1 & \rho_1 & \ddots & & &  & \vdots \\
      0  & \rho_1 & 1 &  \rho_1 & \ddots & &  & \vdots \\
      \vdots & \ddots  & \ddots & \ddots & \ddots & \ddots & &  \vdots \\
      \vdots &  & \ddots  & \ddots & \ddots & \ddots & \ddots & 0  \\
      \vdots &  &  &  \ddots & \rho_1 & 1 & \rho_1 & 0  \\
      \vdots &  & & &  \ddots & \rho_1 & 1 & \rho_1  \\
      0 & \cdots & \cdots & \cdots & \cdots & 0  & \rho_1 & 1
    \end{pmatrix}
  }
\end{eqnarray}


Evaluamos los determinantes. Comenzamos con el numerador.

Expandiendo el numerador por "minors" con respect a la ultima columna, tenemos que el signo es $(-1)^{1+k}$ que
tenemos $\rho_1$ multiplicado por el minor $k-1 \times k-1$ que queda por fuera de las linears rojas resaltadas.
Esa matriz es triangular superior y del determinante de una triangular superior es el producto de los elementos de la diagonal que son todos $\rho_1$ y $k-1$ de ellos.
De forma que el determinate del numerador es $(-1)^{k+1} \rho_1 \rho_1^{k-1}$ es decir

$$ \det \text{numerador} = (-1)^{k+1} \rho_1^{k} \quad (1) . $$

El numerador es una matriz tridiagonal. El determinante de una matriz tridiagonal es un ejercicio interesante,
que desarrollamos en clase.


\begin{eqnarray}
  A_k = \det
  \begin{pmatrix}
    a & b & 0 & 0 & \cdots & 0 \\
    c & a & b & 0 &  \cdots & 0 \\
    0 & c & a & b  &  \cdots & 0 \\
    \vdots & \vdots & \vdots   & \cdots &  \vdots  \\
    0 & 0 &  & \vdots & c & a
  \end{pmatrix} .
\end{eqnarray}
donde $a=1$, $b=c=\rho_1$. 

Ese determinante lo desarrolloamos con respecto a la primera fila que solo tiene 2 elementos no cero, $a,b$.

Expandimos el determinante por la primera fila

$c A_{k-2}$, de forma que escribimos

$$A_k = a A_{k-1} - b( c A_{k-2}) $$


Esta ecuacion es recursiva de orden 2 (en diferencias).
Recuerde la de Fibonacci.

Es homogenea, El polinomio caracteristico 

$$p(\lambda)= \lambda^2 - a \lambda + bc  $$

con raices

$$r_{1,2} = \frac{a \pm \sqrt{a^2 - 4 bc}}{2} $$

Consideremos dos casos:
(podrian ser tres, complejas, reales, repetidas, pero solo
consideramos repetias y los demas)

1. **Raices repetidas**: Esto implica que $a^2 - 4 bc  = 0$. En este caso las raices son $r_1 = r_2 = a/2$. 
La solucion (repasar por favor ecuaciones en diferencias)


$$A_k = (C_1 + C_2 k) \left ( \frac{a}{2}  \right )^k \quad (2) $$

Obviamente, el problema no esta resuelto hasta que no encontremos $C_1, C_2$. Estas se hallan con las condiciones iniciales. Estas son

$A_0=1, A_1=a$. Sera que estas inicialeAs tienen sentido?
Con estas condiciones iniciales quien seria $A_2$

$$A_2 = a A_1 - bc A_0 = a a - bc 1 = a^2 - bc $$

que es el determinante

\begin{eqnarray}
A_2 = \det \begin{pmatrix} a & b \\ c & a\end{pmatrix}
\end{eqnarray}

Con estas condiciones iniciales hallamos $C_1, C_2$.
Tenemos el sistema de la ecuacion 2 para $k=0$

$$ A_0 = (C_1 + C_2 0) \left ( \frac{a}{2}  \right )^0 $$

$$ 1 = C_1$$ o
$$ C_1 = 1$$.

para $k=1$,  volemos a la ecuacion (2).



$A_1 = (1 + C_2) \frac{a}{2}$, con $A_1=a$

$$a = (1 + C_2) \frac{a}{2} $$

\begin{eqnarray}
2a = 1 + C_2 \\
C_2 = 1 \quad  \text{ pues  } a=1
\end{eqnarray}

De forma que $C_1=C_2=1$ y la soucion para raices repetidas es:

$$A_k = (1+k) \left ( \frac{a}{2} \right )^k $$

Tenemos todos los elementos para evaluar $P_k$.

$$P_k = \frac{(-1)^{k+1} \rho_1^k}{(1+k) (1/2)^k} = \frac{-1^{k+1} 2^k \rho_1^k}{1 + k}. \quad (3) $$

Pero sabemos los $\beta$s y no los $\rho$s. Si el discriminante es 0  $a^2 - 4 bc = 1 - 4 \rho_1^2=0$, pues $a=1$, $b=c=\rho_1$.
De forma que $\rho_1=\pm 1/2$. O sea que tenemos:

\begin{eqnarray}
      P_k =  
      \left \{
        \begin{array}{cc}
          \frac{(-1)^{k+1}}{1+k}  &  \text{if } \rho_1=1/2 \\
          -\frac{1}{1+k} & \text{if } \rho_1=-1/2
        \end{array}
      \right . ,
\end{eqnarray}

Si $\rho_1=1/2$, en (3), $2^k \rho_1^k=1$, si $\rho=-1/2$, queda
en el numerador de (3), $(-1)^{k+1}(-1)^{k}=(-1)^{2k+1}= -1 .$

Ahora bien, ustedes para prgramar concen los $\beta$s no los $\rho$s. dejamos como ejercicio probar que si $\rho_1=1/2$ entonces
$\beta_1 = 1$, y que si $\rho_2=-1/2$ entonces $\beta_2=-1$ (asumimos que $\beta_0=1$.

La solucion en termino de los $\beta$s es:


\begin{eqnarray}
      P_k =  
      \left \{
        \begin{array}{cc}

      \right . ,
\end{eqnarray}

2.  **caso 2, raices distintas (complejas o no)**: $a^2 - 4 bc \ne 0$. 
La solucion (recordar ecuaciones en diferencias homogeneas)

$$ A_k = C_1 r_1^k + C_2 r_2^k  \quad (4) $$

De nuevo debemos hallar $C_1, C_2$ de las condiciones iniciales.
Recuerde, estas son $A_0=1$, $A_1=a$, tenmos dos ecuaciones con dos incogintas. De (4)

\begin{eqnarray}
1 &=& C_1 + C_2  \quad \text{para } k=0 \\ 
a  &=& C_1 r_1 + (1 - C_1) r_2  \quad \text{para } k=1
\end{eqnarray}

De la seguna ecuacion despejamos $C_1$,

$$C_1 = \frac{a - r_2}{r_1 - r_2} = \frac{a - (a - \sqrt{a^2 - 4 bc})/2}{2 \sqrt{a^2 - 4bc}/2} = \frac{a+ \sqrt{a^2 - 4 bc}}{2 \sqrt{a^2 - 4bc}}. $$

$$C_2 = 1 - C_1 = 1- \frac{a+ \sqrt{a^2 - 4 bc}}{2 \sqrt{a^2 - 4bc}}= \frac{-a + \sqrt{a^2 - 4 bc}}{2 \sqrt{a^2 - 4 bc}} $$

De forma que la solucion es:

\begin{eqnarray}
A_k &=& C_1 r_1^k + C_2 r_2^k \\
&=& \frac{a+ \sqrt{a^2 - 4 bc}}{2 \sqrt{a^2 - 4bc}} \left ( \frac{a + \sqrt{a^2 - 4 bc}}{2}
\right )^k  + \frac{-a + \sqrt{a^2 - 4 bc}}{2 \sqrt{a^2 - 4 bc}} 
\left ( \frac{a - \sqrt{a^2 - 4 bc}}{2}
\right )^k \\
&=& \frac{1}{\sqrt{a^2 - 4 bc}} \left [
   \left ( \frac{a + \sqrt{a^2 - 4 bc}}{2}
\right )^{k+1} -  \left ( \frac{a - \sqrt{a^2 - 4 bc}}{2}
\right )^{k+1} \right ]
\end{eqnarray}

A mi me interesa esto pero en terminos de $\beta$s. 
Recuerde:
* $\beta_0=1$, $a=1, b=c=\rho_1$ y que de la clase anterior
$\rho_1= \beta_1/(1 + \beta^2)$ 

De forma que

\begin{eqnarray}
a^2 - 4 bc &=& 1 - 4 \rho_1^2 = 1 - 4 \frac{\beta_1^2}{(1 + \beta_1^2)} \\
&=& \frac{(1 + \beta_1^2)^2 - 4 \beta_1^2 }{(1 + \beta_1^2)^2} \\
&=& \frac{1 + 2 \beta_1^2 + \beta_1^4 - 4 \beta_1^2}{(1 + \beta_1^2} \\
&=& \frac{1-2 \beta_1^2 + \beta_1^4}{(1 - 2 \beta_1^2 + \beta_1^4)} \\
&=& \frac{(1 -  \beta_1^2)^2}{(1 + \beta_1^2)^2}
\end{eqnarray}

Vamos a asumir $| \beta_1 | \le 1$, que pasa si $| \beta_1 | >1$?
Si tomamos raiz cuadradas a ambos lados de las expresiones anteriores

$$\sqrt{a^2 - 4 bc} = \frac{1 - \beta_1^2}{1 + \beta_1^2} . $$



\begin{eqnarray}
A_k &=&  \frac{1+ \beta_1^2}{1 - \beta_1^2} \left [
   \left ( \frac{1 +  \frac{1 - \beta_1^2}{1 + \beta_1^2}}{2}
\right )^{k+1} -  \left ( \frac{1 -  \frac{1 - \beta_1^2}{1 + \beta_1^2}}{2}
\right )^{k+1} \right ] \\
&=&  \frac{1+ \beta_1^2}{1 - \beta_1^2} \left [ \left( \frac{1 + \beta_1^2 + 1 - \beta_1^2}{2 (1 + \beta_1^2)} \right )^{k+1} - 
\left ( \frac{1+ \beta_1^2 -1 + \beta_1^2}{2(1 + \beta_1^2)} \right )^{k+1}
    \right ] \\
    &=& \frac{1+ \beta_1^2}{1 - \beta_1^2} \left [ \left ( \frac{2}{2 (1 + \beta_1^2)} \right )^{k+1} -  \left ( \frac{2 \beta_1^2}{2(1 + \beta_1^2)} \right)^{k+1}
        \right ]\\
        &=& \frac{1+ \beta_1^2}{1 - \beta_1^2} \left ( \frac{1}{ (1 + \beta_1^2)^{k+1}} - \frac{ \beta_2^{2k+2}}{(1 + \beta_1^2)^{k+1}} \right ) \\
        &=& \frac{1 + \beta_1^2}{(1 - \beta_1^2) (1 + \beta_1^2)^{k+1}}  (1 - \beta_2^{2k + 2} ) \\
         &=& \frac{1 - \beta_1^{2k + 2}}{ (1 - \beta_1^2)(1 + \beta_1^2)^k}
\end{eqnarray}
donde diga $a=1$.


De forma que

\begin{eqnarray}
P_k = \frac{(-1)^{k+1} \rho_1^k}{A_k} &=&
\frac{(-1)^{k+1} \left [(\beta_1/(1 + \beta_1^2 ) \right ]^k (1 - \beta_1^2) (1 + \beta_1^2)^k}{1 - \beta_1^{2k+2}} \\
&=& \frac{(-1)^{k+1} (1 - \beta_1^2) \beta_1^k}{1 - \beta_1^{2k + 2} }
\end{eqnarray}
