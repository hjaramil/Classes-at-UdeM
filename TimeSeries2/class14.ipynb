## Then complex roots case. Simplification to real numbers implementation

Recuerde $z=a+\mathrm{i} b$ o tambien $z=\rho \cos \theta + \mathrm{i} \rho \sin \theta$, $z = \rho \mathrm{e}^{\mathrm{i} \theta}$ a esta ultima (ecuacion de Euler) se le llama forma **polar**.

Aca

$$ \rho = \sqrt{a^2 + b^2} \quad , \quad \theta = \arctan(b/a) $$

Recuerde que la ecuacion caracteristica del sistem AR(2) es

$$ p(B) = 1 - \phi_1 B - \phi_2 B^2$$

Las raices de este polinomio son (formula cuadratica).

$$ \phi_{\pm} = \frac{\phi_1 \pm \sqrt{ \phi_1^2 + 4 \phi_2}}{-2 \phi_2} $$


Las raices complejas implican que el discriminante es negativo, 
es decir $\phi_1^2 - 4 \phi_2 < 0$, podemos escribir

$$ \sqrt{\phi_1^2 + 4 \phi_2} = \sqrt{-1 (- \phi_1^2 - 4 \phi_2)} = \mathrm{i} \sqrt{-\phi_1^2 - 4 \phi_2} $$

De forma que la solucion a mi ecuacion caracteristica es:

$$ r_{\pm}  =  \frac{\phi_1 \pm \mathrm{i} \sqrt{-\phi_1^2 - 4 \phi_2 }}{-2 \phi_2} $$

Dejo de atera probar que usted puede factorizar el polinomio caracteristico

$$1 - \phi_1 B - \phi_2 B^2 = (1 - \varphi_1 B)(1 - \varphi_2 B)$$.

donde las raices $r_1, r_2$ estan definidas por

$$r_1 = \frac{1}{\varphi_1}  \quad , \quad r_2 = \frac{1}{\varphi_2} $$


$$r_1 = -\frac{\phi_1}{2 \phi_2} + \mathrm{i} \frac{\sqrt{-\phi_1^2 - 4 \phi_2}}{2 \phi_2 } = a + \mathrm{i} b $$

donde


$$ a = -\frac{\phi_1}{2 \phi_2}  \quad , \quad b=  \frac{\sqrt{-\phi_1^2 - 4 \phi_2}}{2 \phi_2 }$$
$a, b \in \mathbb{R}$. 

Se puede probar de igualforma que $r_2 = a - \mathrm{i} b$. Esto
se deja como ejercicio al estudiante.  En forma polar

$$ r_{\pm} = r_{1,2}  = \rho \mathrm{e}^{\pm \mathrm{i} \theta}$$
donde

$$ \rho=\sqrt{a^2 + b^2} \quad , \quad \theta = \arctan \left ( \frac{b}{a} \right )$$
Separadamente

$$r_1 = \rho \mathrm{e}^{\mathrm{i} \theta} \quad , \quad r_2 = \rho \mathrm{e}^{-\mathrm{i} \theta}  $$

El objetivo de este ejercicio es simplificar la expresion (1). Que es la solucion analitica del modelo AR(1). 
Para esta simplificacion asumimos primero que

$$ X_0 = Z_0 \quad , \quad X_1 = Z_1 $$ de forma que

\begin{eqnarray}
 C_1 &=& \frac{-r_1 X_0 - r_2 Z_0 + r_1 r_2 X_1 - r_1 r_2 Z_1 }{r_2 - r_1} =  \frac{-r_1 X_0 - r_2 X_0 + \cancel 
 \{ {r_1 r_2 X_1 - r_1 r_2 X_1 } \} }{r_2 - r_1}  \nonumber \\
 &=& \frac{r_1 + r_2}{r_1 - r_2} X_0\\
      C_2 &=& \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 Z_1 + r_1 Z_0 }{r_2 - r_1} \\
      &=&  \frac{r_2 X_0 - \cancel{ \{ r_1 r_2 X_1 + r_1 r_2 X_1 \}}+ r_1 X_0 }{r_2 - r_1} \\
      &=& - \frac{r_1 + r_2}{r_1 - r_2} X_0 
\end{eqnarray}

Observe que $C_2 = -C_1$

Recuerden $r_1 = \rho \mathrm{e}^{\mathrm{i} \theta} , r_2 = \rho \mathrm{e}^{-\mathrm{i} \theta}$

Del repaso de variable compleja en las notas a mano alzada:

$$ r_1 + r_2 = 2 \rho \cos \theta  \quad , \quad r_1 -r_2 = 2 \mathrm{i} \rho \sin \theta$$

De forma que 


$$C_1 = \frac{2 \rho \cos \theta}{2 \mathrm{i} \rho \sin \theta} X_0 = \frac{\cos \theta}{\mathrm{i} \sin \theta} X_0 = - 
\mathrm{i}\cot \theta X_0$$


$$C_2 = \mathrm{i} \cot \theta X_0$$



\begin{eqnarray}
C_1 \left ( \frac{1}{r_1} \right ) ^n + C_2  \left ( \frac{1}{r_2} \right )^n &=& \\
      &=& X_0(-\mathrm{i} \cot \theta  \rho^{-n} \mathrm{e}^{-\mathrm{i} n \theta} + \mathrm{i} \cot \theta \rho^n \mathrm{e}^{\mathrm{i} n \theta} ) \\
      &=& \mathrm{i} \rho^{-n} \cot \theta X_0 ( \mathrm{e}^{\mathrm{i} n \theta} -\mathrm{e}^{-\mathrm{i} n \theta} ) \\
      &=&  \mathrm{i} \rho^{-n} \cot \theta X_0 (2 \mathrm{i} \sin (n \theta) ) \\
      &=& -2  \rho^{-n} \cot \theta \sin (n \theta) X_0
\end{eqnarray}




\begin{eqnarray}
 && \frac{1}{r_1-r_2} \left [ \left ( r_1 \frac{1}{r_2} - r_2 \frac{1}{r_1}  \right ) 
        Z_{n-1} + \cdots +  \left ( r_1 \left ( \frac{1}{r_2} \right)^n - r_2 \left ( \frac{1}{r_1}  
      \right )^n \right ) Z_0 \right ]   \\
    &=&  \frac{1}{2 \mathrm{i} \rho \sin \theta} \left [ ( \mathrm{e}^{2 \mathrm{i} \theta} - \mathrm{e}^{- 2 \mathrm{i} \theta}) Z_{n-1} + \cdots + \frac{1}{\rho^{n-1}} ( \mathrm{e}^{\mathrm{i} (\theta + n \theta)} - \mathrm{e}^{\mathrm{i}(-\theta - n \theta)}) Z_0
        \right ]  \\
        &=& \frac{1}{2 \mathrm{i} \rho \sin \theta} \left [  2 \mathrm{i} \sin (2 \theta) Z_n + \cdots + \frac{2 \mathbb{i} \sin (n+1)\theta }{\rho_{n-1}}
            \right ] \\
        \frac{1}{ \rho \sin \theta} \left [   \sin (2 \theta) Z_n + \cdots + \frac{\sin (n+1)\theta }{\rho_{n-1}}
            \right ]
\end{eqnarray}

Usando estos resultados tenemos que

\begin{eqnarray}
X_0 &=& Z_0 \\
X_1 &=& Z_1 \\
&\vdots& \\
X_n =  -2  \rho^{-n} \cot \theta \sin (n \theta) X_0 + Z_n + 
\frac{1}{ \rho \sin \theta} \left [   \sin (2 \theta) Z_n + \cdots + \frac{\sin (n+1)\theta }{\rho_{n-1}}
            \right ]
\end{eqnarray}

Esta solucion tiene la vantaja de que solo usa numeros reales.

Todo esto asume que $r_1 \ne r_2$. Dejo como ejercicio derivar (dedudir) las ecuaciones cuando $r_1=r_2$.

## Parametros estadisticos del modelo AR(2).
Recuerden el modelo

$$ X_t = c + \phi_1 X_{t-1} + \phi_2 X_{t-2} + Z_t $$

Vamos a encontrar los parametros estadisticos asumiendo que la serie de tiempo es **estacionaria debil**.


* **Media** (mean):
\begin{eqnarray}
\mu &=&\mathbb{E}[X_t] = \mathbb{E}[c] + \phi_1 \mathbb{E}[X_{t-1}] + \phi_2 \mathbb{E}[X_{t-2}] + \mathbb{E}[Z_t] \\
&=& c + \phi_1 \mu + \phi_2 \mu + 0 
\end{eqnarray}
Despejamos $\mu$.
$$ \mu = \frac{c}{1 - \phi_1  - \phi_2}$$
claramente si no hay "drift" $(c=0)$, la media es 0.

* **Varianza**:
De $X_t = \phi_1 X_{t-1} + \phi_2 X_{t-2} + Z_t$, $c=0$.

"A naive approach", usando propiedades de la varianza y la hipotesis de estacionarieidad debil

$$ \text{Var}{X_t}= \sigma^2$$.

$$ \sigma^2 = \phi_1^2 \sigma^2 + \phi_2^2 \sigma^2 + \sigma_Z + 2 \text{Cov}[]....  $$
Tarea: continuar este ejercicio hasta obtener $\sigma^2$.

Hagamos el problema por otro lado.

\begin{eqnarray}
\gamma_k &=& \mathbb{E}[X_t X_{t-k}] = \phi_1 \mathbb{E}[X_{t-1} X_{t-k}] + \phi_2 \mathbb{E}[X_{t-2} X_{t-k}] + \mathbb{E}[Z_t X_{t-k}] \\
&=& \phi_1 \gamma_{k-1} + \phi_2 \gamma_{k-2} + \mathbb{E}[Z_t X_{t-k}] 
\end{eqnarray}

donde

\begin{eqnarray}
\mathbb{E}[Z_t X_{t-k}]  = \left \{
    \begin{array}{cc}
    \sigma_Z^2 & \text{si  } k = 0 \\
    0 & \text{si } k=1,2, ,\cdots
    \end{array}
    \right .
\end{eqnarray}

Recuerden que $\gamma_t = \gamma_{-t}$ (simetrica).


Cominemos estas dos ultimas ecuaciones.

\begin{eqnarray}
\sigma^2 &=& \gamma_0 = \phi_1 \gamma_1 + \phi_2 \gamma_2 + \sigma_Z^2  \quad , \quad \text{recuerde que } k=0 \\
\gamma_k &=& \phi_1 \gamma_{k-1} + \phi_2 \gamma_{k-2} \quad , \quad \text{para } k=1,2, 
\end{eqnarray}
 
 Para $k=1$,

 $$\gamma_1 = \phi_1 \gamma_0 + \phi_2 \gamma_1 $$

 despejemos $\gamma_1$,

 $$ \gamma_1 =  \frac{\phi_1 \gamma_0}{1-\phi_2} =\frac{\phi_1 \sigma^2}{1-\phi_2} $$

 Para $k=2$.
 

Recuerden que $\gamma_t = \gamma_{-t}$ (simetrica).


Cominemos estas dos ultimas ecuaciones.

\begin{eqnarray}
\sigma^2 &=& \gamma_0 = \phi_1 \gamma_1 + \phi_2 \gamma_2 + \sigma_Z^2  \quad , \quad \text{recuerde que } k=0 \\
\gamma_k &=& \phi_1 \gamma_{k-1} + \phi_2 \gamma_{k-2} \quad , \quad \text{para } k=1,2, 
\end{eqnarray}
 
 Para $k=1$,
