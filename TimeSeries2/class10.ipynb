# El Backshift operator (operador de deplazamiento hacia atras)
El objeto de este es simplificar.

$$ B(X_n) = X_{n-1}$$

Por decir algo $X=(1,2,3)$
$$B(X_0) = 0, B(X_1)= 1, B(X_2)=2 $$

$B^0=I$.

Por induccion.

$$B^k(X_n)= B^{k-1} B(X_n) = B^{k-1} X_{n-1} = \cdots = B^{k-i} X_{n-i} = X_{n-k}   $$
De ahora en adelante vamos a usar mucho este operador.

Comenzamos con el modelo AR(1):

\begin{eqnarray}
 X_t &=& \phi X_{t-1} + Z_t +c \quad  \text{         usamos $c$=0}\\
 X_t &=& \phi B X_t + Z_t \\
 X_t - B \phi X_t &=& Z_t \\
 (1 - \phi B)X_t &=& Z_t 
\end{eqnarray}

Postulamos lo siguiente:

$$X_t = \sum_{j=0}^\infty \varphi_j Z_{t-j} = \varphi_0 Z_t + \varphi_1 Z_{t-1} + \cdots = \phi_0 B^0 Z_t + \phi_1 B^1 Z_t + \cdots = \varphi(B) Z_t  $$
donde $\varphi(B) = \sum_{j=0}^\infty \varphi_j B^j .$
El objetivo ahora es hallar los coeficientes $\varphi_i$.


$$X_t = \varphi(B) Z_t  = \varphi(B) (1 - \phi B)X_t $$

De aca que

\begin{eqnarray}
1 X_t &=& (\varphi_0 + \varphi_1 B^1 + \varphi_2 B^2 + \cdots + \varphi_j B^j + \cdots +)(1 - \phi B) X_t \\
1 &=& (\varphi_0 + \varphi_1 B^1 + \varphi_2 B^2 + \cdots + \varphi_j B^j + \cdots +)(1 - \phi B)  \\
1 &=& \varphi_0 + \varphi_1 B^1 + \varphi_2 B^2 + \cdots + \varphi_j B^j + \cdots + \cdots - \varphi_0 \phi  B - \varphi_1 \phi B^2 - \varphi_2 \phi B^3 - \cdots - \varphi_{j-1} \phi B^j \cdots \\
&=&   \varphi_0 + (\varphi_1 - \varphi_0 \phi)B + (\varphi_2 - \varphi_1 \phi) B^2 + \cdots (\varphi_j - \varphi_{j-1} \phi)B^j + \cdots 
\end{eqnarray}

"matching coefficients".


\begin{eqnarray}
1 &=& \varphi_0 \\
\varphi_1 &=& \varphi_0 \phi = \phi \\
\varphi_2 &=& \varphi_1 \phi = \phi \phi = \phi^2 \\
&\vdots& \\
\varphi_j &=& \varphi_{j-1} \phi = \phi^{j-1} \phi = \phi^j
\end{eqnarray}

Des decir $\varphi^j = \phi^j$ , $j=0,1,2, \cdots $
Reemplazamos arriba y encontramos.

$$X_t = \sum_{j=0}^\infty \varphi_j Z_{t-j} =\sum_{j=0}^\infty \phi^j Z_{t-j} $$

Yo quiero que pensemos en los opeardores como numeros que no conmutan.

$$(1 - \phi B)(1 + \phi B + \phi B^2 + \cdots + \phi^{n-1} B^{n-1}) = 1 - \phi^n B^n $$

Algebra de operadores (multiplicacion , distribucion), y agrupar terminos).

Entonces podemos pensar que

$$ \frac{1 - \phi^n B^n}{1 - \phi B} = \sum_{i=0}^{n-1} \phi^i B^i $$

**ojo** los operadores no se dividen asi. Piense que $B$ es una matriz
que es $1/(1 - \phi B)$ ? 

Sin embargo pensamoes en $1 - \phi B$ como $I - \phi B$ y que si el operador
tiene inverso $1/(1 - \phi B) = (I - \phi B)^{-1}$. 

Si $| \phi | < 1$ podemos tomar el limite al $\infty$.

$$ \frac{1 }{1 - \phi B} = \sum_{i=0}^\infty \phi^i B^i  \quad (5) $$

La serie de potencias

$$\frac{1}{1-x} = 1 + x + x^2 + \cdots x_n + \cdots  $$
y esta serie de potencias converge si $|x| < 1$. 
[La serie se llama geometrica](https://en.wikipedia.org/wiki/Geometric_series).

DE forma que

$$ X_t = \frac{1}{1 - \phi B} Z_t = \sum_{j=0}^{\infty} \phi^j B^j Z_t =\sum_{j=0}^{\infty} \phi^j Z_{t-j}.  \quad (6) $$

Otra razon mas para pensar que $| \phi | < 1$. Luego hablamos mas de estas restricciones.





# Linear Difference Equations: Ecuaciones en diferencias lineales
no confundir con Linear differential equations (ecuaciones diferenciales lineales).
Ver en [Wikipedia](https://en.wikipedia.org/wiki/Linear_differential_equation).


Las ecuaciones diferenciales son para el continuo y las "en diferencias" son para el discreto. 

La forma general de una ecuacion en diferencias lineal de orden $n$ es:

$$ a_0 x_t + a_1 x_{t-1} + \cdots a_n x_{t-n} = f_t  \quad (1) $$
donde $a_i, i=0,1, \cdots, n$ son coeficientes (mas adelante van a ser nuestras incoginas), los $x_t$ son terminos de la serie de $x$. $f_t$ se llama "forcing term" (termino de fuerza, o fuente).

Mas adelante vamos a ver que los coeficientes $a_i$ se pueden determinar usando lo que llaman Yule-Walker equations. 

Hay una relacion entre ecuaciones en diferencias y ecuaciones diferenciales, que para el interesado la encuentra 
[aca](https://drive.google.com/file/d/1PlD4bLGwvBJKyQm8FniM7v5N_ZbOg_97/view).

Vamos a ver como nuestros problemas AR(P) (solo hasta este momento conocemos AR(1)) son una consecuencia de las ecuaciones en diferencias.

Comencemos con el modelo AR(1)

$$ X_t = c + \phi X_{t-1} + Z_t$$
Pasando a restar el termino en $X_{t-1}$

$$X_t - \phi X_{t-1} = c + Z_t $$

Observe que esta ecuacion coincide con la ecuacion en diferencias (1) teniendo en cuenta que

* $X_t \leftarrow x_t$
* $a_0 = 1$
* $a_1 = -\phi$
* $f_t = c + Z_t$

Por "cuarta" vez vamos a encontrar la solucion de $X_t$ asumiendo que conocemos los $a_i$ (note que otro problema es conocer los $a_i$ teniendo los $X_t$).
Al primero lo podemos llamar modelamiento y al segundo inversion (estimacion).

De igual forma que en las ecuaciones diferenciales existe una solucion particular y una solucion homogenea y que de ambas se construye LA SOLUCION, 
en ecuaciones en diferencias tambien tenemos una solucion particular y una homogenea y la combinacion de ambas con condiciones iniciales nos produce
LA SOLUCION unica.

## Problema homogeneo.
Homogeneo quiere decir $f_t=0$ en la ecuacion (1)

En este momento la incoginta es $X_t$, mas adelante
(Yule-Walker) la incognita es $\phi$. Asumimos entonces
que conocemos a $\phi$. 


$$X_t - \phi X_{t-1} = 0  $$
Se puede ver de la propagacion hacia adelante que

\begin{eqnarray}
X_1 &=& \phi X_0 \\
X_2 &=& \phi X_1 = \phi \phi X_0 = \phi^2 X_0 \\
&\vdots& \\
X_t &=& \phi^t X_0 
\end{eqnarray}

Vamos a usar, de nuevo, el backshift operator para resolver el problema.

$$ (1 - \phi B) X_t = 0  \quad  (2) $$. Al polinomio


$$ p(B) = 1 - \phi B $$
se le llama **polinomio caracteristic**. 

Parentesis, comparacion con las ecuaciones diferenciales lineales, ordinarias
(
$y'' + a y + 5 = 7$,  $r^2 + ar + 5 = 0$, las soluciones son
de la forma $A \exp(r_1) + B \exp(r_2) $
)

Se resuelve la ecuacion caracteristica $p(B)=0$, para $B$:

$$ r = \frac{1}{\phi} \quad (3) $$. De la teoria de ecuciones en diferencias, lineales se tiene que la solucion al problema homogeneo (2) es

$$X_{th} = C \left ( \frac{1}{r} \right)^t = C \phi^t  \quad (3) $$
donde $X_{th}$ es una serie de tiempo, solucion homogenea. 

Verifiquemos que esta  es una solucion general (homogenea) al problema.

Recuerde que $\phi= 1/r$ de (4)
$$(1 - \phi B) X_{th} = X_{th} - \phi X_{th-1} = C \left( \frac1r \right)^t - \phi C \left ( \frac1r \right )^{t-1} = C \left (\frac1r \right ) - C \left (\frac1r \right )^t = 0  $$

Hay infinidad de soluciones por que $C \in \mathbb{R}$ es cualquier real.
Para poder fijar la solucion necesitamos (como en las ecuaciones diferenciales)
una condicion inicial. Asumimos que $X_t=X_0$ para $t=0$ conocida.

En nuestra solucion tenemosa de la ecuacion (3)

$$X_{th=0} = C \left ( \frac{1}{r} \right ) ^0 = X_0  $$
de forma que $C=X_0$. Y la solucion homogenea es

$$ X_{th} = \left ( \frac{1}{r} \right)^t X_0 = X_0 \phi^t $$

Ahora bien falta el problema de la solucion particular.  Es decir, 
econtrar $X_{pt}$ tal que

$$(1 - \phi B) X_t = c+ Z_t $$
donde $f_t = c + Z_t$. 

Para hallar esta solucion particular usamos la aproximacion.

$$ \frac{1 }{1 - \phi B} = \sum_{i=0}^\infty \phi^i B^i  $$


La solucion particular $X_{tp}$ esta dada por:

$$ X_{tp} = (1- \phi B)^{-1} (c + Z_t) = (1 + \phi B + \phi^2 B^2 + \cdots + \phi^n B^n + \cdots ) ( c  + Z_t)$$.
Lo mismo que en ecuaciones diferenciales la solucion general es la suma de la particular mas la homogenea. $X_t = X_{th} + X_{tp}$.

Resultado preliminar:

$$(1 - \phi B) X_t = (1 - \phi B)( X_{th} + X_{tp}) = \cancel{(1 - \phi B) X_{th}} + (1 - \phi B) X_{tp} = (1 - \phi B) X_{tp} \quad (7) $$

Escribimos 

\begin{eqnarray}
X_t &=& X_{th}  + X_{tp} \\
&=& \phi^t C + c(1 + \phi B + \phi^2 B^2 + \cdots + \phi^n B^n + \cdots ) + 
Z_t + \phi Z_{t-1} + \cdots + \phi^n Z_{t-n} + \cdots \\
&=& C \phi^t + c \sum_{i=0}^\infty \phi^i + \sum_{i=0}^\infty \phi^i Z_{t-i} 
\end{eqnarray}

Si aplicamos las condiciones iniciales $X_t=X_0$, $t=0$,
entonces $C=X_0$.

$$ X_t = X_0 \phi^t + c \sum_{i=0}^\infty \phi^i + \sum_{i=0}^\infty \phi^i Z_{t-i} $$
