# Time Series Models : Modelos de series de tiempo
## AR(P)
## p$=1$


Vamos a estudiar modelos de autoregresion.

En DSP: Digital Signal Processing (procesamiento de datos digitales)

El proceso mas importante en DSP es la convolucion.

$$ y_i = \sum_{k=1}^p x_{i-k} h_k = (x * h)_i $$
Aca los datos son $x_i$, el filtro es $h$ (es otra serie de tiempo) la salida es $y$. A este proceso se le llama **convolucion** y es un proceso lineal.

Hay un caso de esto muy especial y es cuando $y_j$ es la misma $x_i$ pero
con un "delay", es decir $y_i = x_{i+j}$.  Es decir:

$$ y_i = x_{i+j} = \sum_{k=1}^p x_{i-k} h_k = (x * h)_i $$
A este modelo se le llama **predictivo**. En el caso de DSP se le llama
convolucion predictiva. Este es el modelo AR(p) que vamos a estudiar en este curso. 

Si nosotros quisieramos obenter el filtro  $h$ de los datos $x$ y el proceso
convolutivo, entonces estamos haciendo un proceso que en

* series de tiempo: es estimacion (estimamos h), llegamos a las ecuaciones de
 **Yule-Walker**. El proceso de modelamiento (obtener $y$ de $x$ y el filtro $h)$) se llama procesos autorgresivos. Lo que estudiamos aca.
* DSP: inversion, en este caso estamos hablando de **predictive deconvolution**, en particular se estudian los **Wiener Filters**.
[ver este link](https://wiki.seg.org/wiki/Predictive_deconvolution).

Es el mismo problema pero visto con distintos nombres en dos comunidades
diferentes.

En el caso de que $j=0$.
$$x_{i} = \sum_{k=1}^p x_{i-k} h_k = (x * h)_i $$

En la practiva obtenido un filtro $h$ no es posible obtener precision infinita en el modelo de la ecuacion anterior. Entonces existe un error
y podemos escribir mejor:

$$x_{i} = \sum_{k=1}^p x_{i-k} h_k + \epsilon_i$$


Vamos a particularizar el proceso hacia lo que se llama procesos de autoregresion AR(p).
Usando la notacion de series de tiempo escribimos el proceso como

$$X_t = c + \sum_{k=1}^p \phi_k X_{t-k} + Z_t $$
donde ahora:

* Introdujimos $c$ como un "bias" sesgo, que muchas veces en este curso es $c=0$. 
* Cambiamos $x_i$ for $X_t$, para ajustarnos a la notacion de series de tiempo. Aca $t$ es el subinidice sobre los enteror.
* $Z_t$ es el ruido $\epsilon_t$, es ruido blanco.

Esta ecuacion nos va ocupar la mayoria de este curso. 

Comenzamos con  el caso mas simple p$=1$ (si p$=0$ solo tenemos $x=c+Z$ que no
es interesante).  Es decir

$$ X_t = c + \phi_1 X_{t-1} + Z_t.$$

Un caso especial de este proceso llamado AR(1), es cuando $c=0$, $\phi_1=1$ y se llama **random walk** (movimiento Browniano, o de borracho).

$$ X_t = X_{t-1} + Z_t . $$


## Proceso AR(1)

$$X_t = c + \phi_1 X_{t-1} + Z_t \quad (*) $$
a $c$ se le llamma **dritf**. Iniciamos el proceso con $X_0=c+ Z_0$ de
forma que $X_1 = c + \phi_1 X_0 + Z_1 = c + \phi_1 ( (c + Z_0) + Z_1)$

En el proceso AR(1) voy a llamar a $\phi_1$ como $\phi$, por que no es sino uno.

Vamos a estudiar los parametros estadisticos: media, varianza, ACF (Auto Correlation Function). 

###  **Media**: 

\begin{eqnarray}
\mu = \mathbb{E}(X_t) &=& \mathbb{E}(c + \phi X_{t-1} + Z_t) \\
&=& \mathbb{E}(c) + \phi \mathbb{E}(X_{t-1}) + \mathbb{E}(Z_t) \\
&=& c + \phi \mu
\end{eqnarray}
Decimos que $\mathbb{E}(X_{t-1}) = \mathbb{E}(X_t)=\mu$ por que vamos
a asumir que el proceso es estacionario debil. 

Despejando $\mu$ tenemos que

$$\mu = \frac{c}{1-\phi} . \quad (**)$$ 

Notas: 

* Si $\phi=1$ tenemos problemas. Vamos a ver mas adelante que para
que este proceso este bien definido $|\phi | < 1$. 
* Si $c=0$, entonces $\mu=0$. Claramente la media no depende de $t$, de la ecuacion (**) pero ojo, eso ya lo habiamos asumido. 

Vamos a buscar una manera de no tener que usar el drift $c$ en lo que sigue.

Hacemos un cambio de variables en la ecuacion (*). 

$$X_t \leftarrow X_t - \mu \quad  (***) $$. 

Inicialmente usamos la ecuacion (*) y reemplazamos $c=\mu(1 - \phi)$ de la ecuacion (**).


$$X_t =  \mu(1 - \phi) + \phi X_{t-1} + Z_t = \phi(X_{t-1} - \mu) + \mu + Z_t $$

pasamos $\mu$ a la izquierda y obenemos

$$X_t - \mu =  \phi(X_{t-1} - \mu) + Z_t. $$
Ahora si hacemos el cambio de variable sugerido por la ecuacion (***), 
y obtenemos

$$X_t = \phi X_{t-1} + Z_t$$

Lo que estamos diciendo es que si le removemos la media a los datos antes de comenzar a trabajar, el termino $c$ de la regresion AR(1) no desaparece.
Eso simplifica los procesos. Ahora bien podemos recuperarlo al final. 
Es decir

* Se remueve la media de los datos (incomplete mean normalization $X_t \leftarrow X_t - \mu$)

* Se analiza la serie AR(1) donde $c$ ya no existe. 

* Al resultado de todo anterior se le agrega el "drift" c.

### Varianza:
Asumimos, por simplicidad $c=0$. 

Es sabido que

* $$\mathbb{V}[X+Y] = \mathbb{V}[X] + 2 \text{Cov}[XY] + \mathbb{V}[Y]  $$
y 
*   $$\mathbb{V}[c X] = c^2 \mathbb{V}[X]$$. 
\begin{eqnarray}
\sigma^2 &=& \mathbb{V}(X_t) = \mathbb{V}[\phi X_{t-1}] + 2 \text{Cov}[\phi X_{t-1} Z_t] + \mathbb{V}(Z_t) 
\end{eqnarray}

* De tarea dejo que el estudiante pruebe
$$ \text{Cov}[ \phi X_{t-1} Z_t] = 0$$.
* Llamamos $\sigma_t^2 = \mathbb{V}[Z_t]$.
* $\mathbb{V}(\phi X_{t-1}) = \phi^2 \mathbb{V}[X_{t-1}] = \phi^2 \sigma^2$.
Asumiendo estacionariedad debil. (se converva la varianza con el tiempo).

Entonces

$$ \sigma^2 = \phi^2 \sigma^2 + \sigma_z^2$$
Despejamos $\sigma^2$ y obtenemos

$$ \sigma^2 = \frac{\sigma_Z^2}{1 - \phi^2}$$

Esta es la varianza de un proceso AR(1).
De nuevo si $\phi= \pm 1$ tenemos problemas.
Si expande $1/1-\phi^2$ en una serie geometrica $1 + \phi + \phi^2 + \cdots $,
es necesario que $ | \phi | \le 1$. Sobre esto insistimos de nuevo mas adelante.

### Funcion de autocorrelacion ACF.

Recordemos de la clase anterior

$$\gamma_1 = \text{Cov} (X_t X_{t+1}) = \mathbb{E}[ X_t (\phi X_t + Z)] = \mathbb{E}[ \phi X_t^2 + X_t  Z_{t+1}] = \phi \mathbb{E} [X_t^2] + \mathbb{E}[X_t Z_{t-1}] $$
El ultimo termino es 0. Esto queda de tarea.

De forma que

$$\gamma_1 = \phi \sigma^2. $$

Movamonos ahora a ordenes superiores.

\begin{eqnarray}
X_{t+ \tau} &=& \phi X_{t + \tau -1} + Z_{t + \tau} \\
X_t X_{t+ \tau} &=& \phi X_t X_{t + \tau -1} + X_t Z_{t + \tau}  \quad \text{multiplicando a izquierda por $X_t$}\\
\mathbb{E}(X_t X_{t+ \tau}) &=& \mathbb{E}(\phi X_t X_{t + \tau -1}) + \mathbb{E}(X_t Z_{t + \tau})  \quad \text{tomando la esperanza}\\
\end{eqnarray}
De nuevo, el estudiante debe probar que $\mathbb{E}(X_t Z_{t + \tau}) = 0$.

Entonces

$$\gamma_{\tau} = \phi \gamma_{\tau-1} . $$

Tengo la siguiente recursion

\begin{eqnarray}
\gamma_0  &=&\sigma^2 \\
\gamma_1 &=& \phi \sigma^2 \\
\gamma_2 &=& \phi \gamma_1 = \phi^2 \sigma^2 \\
&\vdots & \\
\gamma_{\tau} &=& \sigma^2 \phi^{\tau}
\end{eqnarray}


La ACF se encuentra normalizando (dividiendo por $\gamma_0=\sigma^2$. Es decir


\begin{eqnarray}
\rho_0  &=&1  \\
\rho_1 &=& \phi  \\
\rho_2 &=&  \phi^2  \\
&\vdots & \\
\rho_{\tau} &=&  \phi^{\tau}
\end{eqnarray}

La ACF es $\rho_{\tau} = \phi^{\tau}$. 


### Otra forma de encontar las propiedades del modelo AR(1).

Recuerde la ecuacion clave

$$X_t = \phi X_{t-1} + Z_t $$

#### Propagacion hacia atras (backpropagation).

\begin{eqnarray}
X_t &=& \phi X_{t-1} + Z_t \\
X_t &=& \phi ( \phi X_{t-2} + Z_{t-1}) + Z_t = \phi^2 X_{t-2} + \phi Z_{t-1} + Z_t \\
X_t &=& \phi^2 (\phi X_{t-3} + Z_{t-2}) + \phi Z_{t-1} + Z_t = \phi^3 X_{t-3} + \phi^2 Z_{t-2} + \phi Z_{t-1} + Z_t \\
&\vdots& \\
X_t &=& \phi^k X_{t-k} + \sum_{j=0}^{k-1} \phi^j Z_{t-j}. \quad (v)
\end{eqnarray}

Note que esto tiene la forma de una convolucion.

La pregunta es puede $k \to infty$? y que se necesita?

Veamos, escribimos a ultima ecuacion de la forma

$$ \phi^k X_{t-k} = X_t - \sum_{j=0}^{k-1} \phi^j Z_{t-j} \quad (****)$$

Voy a tomar la varianza y el limite cuando $k \to \infty$.

$$ \lim_{k \to \infty} \mathbb{E} \left [ \left (X_t - \sum_{j=0}^{k-1} \phi^j Z_{t-j}
    \right )^2
    \right ] = \lim_{k \to \infty}  \phi^{2k} \mathbb{E}[ X_{t-k}]^2 $$

donde el termino de la derecha viene del de la izquierda en la ecuacion (****)

Aca es donde vemos que $| \phi | < 1$ por que sino, el termino de la derecha
de la ecuacion anterior se va al infinito. 

Tomamos la ecuacion (v) y el limite cuando $k \to \infty$, el primero termino
desaparece y encontramos

$$X_t = \sum_{j=0}^{\infty} \phi^j Z_{t-j}  $$

Dejamos como tarea encontar la media, varianza y ACF de esta serie de tiempo expresada de esta forma.

Pensemos en la propagacion hacia adelante.

\begin{eqnarray}
X_0 &=& Z_0 \\
X_1 &=& \phi X_0 + Z_1 = \phi Z_1 + Z_0 \\
X_2 &=& \phi X_1 + Z_2 = \phi( \phi X_0 + Z_1) + Z_2 = \phi^2 X_0 + \phi X_1 + Z_2 
\end{eqnarray}
y asi suscesivamente.
