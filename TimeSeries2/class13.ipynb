# Sistema homogeneo 
Adelantamos un poco.
No forcing term $Z_t + c$.
$$X_t = \phi_1 X_{t-1} + \phi_2 X_{t-2} $$
cuando pasa todo a la izquierda obtiene

$$X_t - \phi_1 X_{t-1} - \phi_2 X_{t-2} = 0 $$
en terminos del backshift $B$

$$(1 - \phi_1 B - \phi_2 B^2) X_{th} = 0 $$


## no repetidas. Diferentes
La ecuacion caracteristica es
 
$$p(B) = 1 - \phi_1 B - \phi_2 B^2 = (1 - \varphi_1 B)(1 - \varphi_2 B) = 0 $$ -
donde las raices **distintas** son $r_1 = 1/\varphi_1$, $r_2=1/\varphi_2$.

$$p(B) X_{th} = (1 - \phi_1 B - \phi_2 B^2) X_{th} = \left (1 - \frac{B}{r_1} \right ) \left (1 - \frac{B}{r_2} \right ) X_{th} = 0 $$


Este sistema es relativamente facil de resolver por que si $ab=0$, $a=0$, o $b=0$.

Es decir tenemos dos ecuaciones de primer orden

\begin{eqnarray}
(1 - B/r_1) X_{th} &=& 0 \\
(1 - B/r_2) X_{th} &=& 0 
\end{eqnarray}

Las dos soluciones (de la primera y segunda ecuacion.

\begin{eqnarray}
X_{th0} &=& C_1 \left (   \frac{1}{r_1}  \right )^t \\
X_{th1} &=& C_2 \left (   \frac{1}{r_2}  \right )^t \\
\end{eqnarray}

Si solo quisieramos resolver el sistema homogeneo, debemos encontrar $C_1, C_2$.

La solucion al sistema homogeneo es una combinacion de las dos anteriores

$$X_{th} = C_1 \left ( \frac{1}{r_1} \right )^t + C_2 \left ( \frac{1}{r_2} \right )^t $$

Cualquier combinacion con  $C_1, C_2$ arbitrarios es solucion. Si queremos hallar la unica solucion  para condiciones iniciales dadas debemos pensar de la siguiente forma. Las dos soluciones iniciales son $X_0, X_1$ conocidas para $t=0,1$ respectivamente. Con esto construimos dos ecuaciones con dos incognitas y hallamos $C_1, C_2$.


\begin{eqnarray}
X_0 &=& C_1 + C_2 \quad , \quad t=0 \\
X_1 &=& \frac{C_1}{r_1} + \frac{C_2}{r_2} \quad , \quad t=1
\end{eqnarray}
De la primera ecuacion $C_1 = X_0 - C_2$. Esta expresion la ponenemos en la segunda ecuacion y eliminamos $C_1$. es decir

\begin{eqnarray}
X_1 &=& \frac{X_0 - C_2}{r_1} + \frac{C_2}{r_2} = \frac{X_0}{r_1} + C_2 \left ( \frac{1}{r_2} - \frac{1}{r_1} \right ) \\
&=& \frac{X_0}{r_1} + C_2 \frac{r_1 - r_2}{r_1 r_2} 
\end{eqnarray}
De aca ya despejamos $C_2$.

\begin{eqnarray}
C_2 &=& \left ( X_1 - \frac{X_0}{r_1} \right ) \frac{r_1 r_2}{r_1 - r_2} = \left (  \frac{r_1 X_1 - X_0}{\cancel{r_1}} \right )\frac{\cancel{r_1} r_2}{r_1 - r_2} \\
&=&  \frac{r_2 ( r_1 X_1 - X_0 )}{r_1 - r_2}
\end{eqnarray}
entonces


$$C_1 = X_0 - C_2 = X_0 -   \frac{r_2 ( r_1 X_1 - X_0 )}{r_1 - r_2} = \frac{r_1 X_0 - \cancel{r_2 X_0} - r_1 r_2 X_1 - \cancel{r_2 X_0 }}{r_1 - r_2} = r_1 \frac{X_0 - r_2 X_1}{r_1 - r_2} $$

La solucion del problema es:

\begin{eqnarray}
X_{th} &=& C_1 \left ( \frac{1}{r_1} \right )^t + C_2 \left ( \frac{1}{r_2} \right )^t \\
C_1 &=&  r_1 \frac{X_0 - r_2 X_1}{r_1 - r_2}  \quad , \quad C_2 = \frac{r_2 ( r_1 X_1 - X_0 )}{r_1 - r_2}  \quad , \quad (*****)
\end{eqnarray}

**Ejemplo**:
Existe una sucesion famosa llamada la sucesion de Fibonacci. Esta cumple las siguientes caracteristicas

\begin{eqnarray}
X_0 &=&0  \quad , \quad X_t = 1 \\
X_t &=& X_{t-1} + X_{t-2} \quad, \quad t=2,3, \cdots. 
\end{eqnarray}

## Raices repetidas: Que pasa si las raices son repetitas.
Aca $r_1 = r_2$.  Inicialmente pensemos en un problema mas simple

$$ (1 - B)^2 X_t = 0  \quad (1) .$$  donde $\varphi_1 = \varphi_2 =1$ (caso especial).
El polinomio caracteristico es $p(B)= (1-B)^2$ con la raiz repetida (doble) $r_1=r_2=1$.  La dos soluciones de este problema homogeneo son

$$X_t = C_j t^j $$ donde $j=0,1$, $C_j$ son constantes. Es decir,

\begin{eqnarray}
X_{th0} &=& C_0 t^0 = C_0  \quad , \quad j=0 \quad (2) \\
X_{th1} &=& C_1 t^1 = C_1 t  \quad , \quad j=1  \quad (3)
\end{eqnarray}

Verificamos. La primera solucion

$$(1 -B)^2 C_0 (1-B)(1-B) C_0 = (1-B) (C_0 - C_0) = (1-B)0 = 0$$
lo prueba que es solucion de (1).

Para la segunda

$$(1-B)^2 C_1 t = (1 -B)(1-B) C_1 t = (1-B) (C_1 t -C_1(t-1)) = (1-B)(C_1 t - C_1 t + C_1) =(1-B)C_1 = C_1 - C_1 = 0$$
lo que prueba la segunda solucion.  Queda como tarea al estudiante probar
que $Bc = c$. 

De las condiciones iniciales tenemos

\begin{eqnarray}
C_0 &=& X_0 \quad \text{de la ecuacion  } 2 \\
C_1 &=& X_1 \quad \text{de la ecuacion  } 3.
\end{eqnarray}

Ahora si, pensemos en el problema mas general. Es decir:

$$(1 - \varphi B)^2 X_t = 0 $$

De nuevo se dice que las soluciones son

$X_t= t^j \varphi^t$, $j=0,1$ son las soluciones. 

Tambien verificamos que estas son las dos soluciones.


* Si $j=0$:

    $$(1- \varphi B)^2 X_t = (1 - \varphi B)^2 \varphi^t = (1 - \varphi B)(1 - \varphi B) \varphi^t = (1 - \varphi B)( \varphi^t - \varphi \varphi^{t-1}) = (1 - \varphi B)(\varphi^t - \varphi^t) = (1 - \varphi B)0=0$$

* Si $j=1$:

    \begin{eqnarray}
        (1 - \varphi B)^2 (t \varphi^t) &=& (1 - \varphi B)(1 - \varphi B)(t \varphi^t) \\
        &=& (1 - \varphi B)(t \varphi^t - \varphi (t-1) \varphi^{t-1})  \\
        &=& (1 - \varphi B)(\cancel{t \varphi^t} - \cancel{t \varphi^t} + \varphi^t) \\
        &=& (1 - \varphi B)(\varphi^t) \\
        &=& \varphi^t - \varphi \varphi^{t-1} \\
        &=& \varphi^t - \varphi^t = 0
    \end{eqnarray}

Lo que verifica las dos soluciones.

Estas son

\begin{eqnarray}
X_{th0} &=& C_0 \varphi^t \\
X_{th1} &=& C_1 t \varphi^t
\end{eqnarray}

Como siempre la solucion general es la homogenea mas la particular.

$$X_t = X_{th} + X_{tp} $$







## Solucion al sistema:

### Raices distintas $r_1  \ne r_2$.

\begin{eqnarray*}
      X_t &=&   C_1 \left ( \frac{1}{r_1} \right )^t + C_2   
       \left ( \frac{1}{r_2} \right )^t \\  && 
      + \frac{1}{r_1 - r_2} 
      \left [ (r_1 - r_2) Z_t +  \left (r_1 \frac{1}{r_2} - 
        r_2 \frac{1}{r_1} \right ) Z_{t-1} \right .   \\
        && \left . +  \left (r_1 \frac{1}{r_2^2} - r_2 \frac{1}{r_1^2} \right ) Z_{t-2}  
        \cdots + 
        \left (r_1 \frac{1}{r_2^n} - r_2 \frac{1}{r_1^n} \right ) Z_{t-n} + \cdots 
      \right ]
    \end{eqnarray*}


Toca hallar de nuevo $C_1, C_2$.


Para esto asumimos que conocemos $X_0, X_1$. Asumimos causalidad
$Z_t=0 , t<0$. 


\begin{eqnarray}
X_0 &=& C_1 + C_2 + Z_0 \quad (4) \\
X_1 &=& \frac{C_1}{r_1} + \frac{C_2}{r_2}  + Z_1 + \frac{1}{r_1-r_2} \left ( \frac{r_1}{r_2} - \frac{r_2}{r_1} 
    \right ) Z_0 = \frac{C_1}{r_1} + \frac{C_2}{r_2}  + Z_1 + \frac{1}{r_1-r_2} \frac{r_1^2 - r_2^2}{r_1 r_2} =\frac{C_1}{r_1} + \frac{C_2}{r_2}  + Z_1 + \frac{r_1 + r_2}{r_1 r_2} Z_0
\end{eqnarray}

Para hallar $C_0, C_1$ eliminemos $C_1$, como?, multiplicamos la segunda por $r_1$ y se la restamos a la primera



\begin{eqnarray}
X_0 - r_1 X_1 &=& C_1 + C_2 + Z_0 - C_1 - \frac{r_1}{r_2} C_2 - r_1 Z_1 - \frac{r_1 + r_2}{r_2} Z_0 \\
&=& C_2 + Z_0 - \frac{r_1}{r_2} C_2 - r_1 Z_1 - \frac{r_1 + r_2}{r_2} Z_0 \\
C_2 &=& \frac{X_0 - r_1 X_1 - Z_0 + r_1 Z_1 + \frac{r_1+r_2}{r_2} Z_0}{1 - r_1/r_2 } \\
&=& \frac{r_2 X_0 - r_1 r_2 X_1 - r_2 Z_0 + r_1 r_2 Z_1 + r_1 Z_0 + r_2 Z_0 }{r_2 - r_1}  \\
&=& \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 Z_1 + r_1 Z_0}{r_2 - r_1}
\end{eqnarray}

Ahora de $C_1 = X_0 - Z_0 - C_2$ de la (4)

\begin{eqnarray}
C_1 = X_0 - Z_0 - C_2 &=& X_0 - Z_0 - \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 Z_1 + r_1 Z_0}{r_2 - r_1} \\
&=&  \frac{\cancel{r_2 X_0} - r_2 Z_0 - r_1 X_0 +  \cancel{r_1 Z_0} - \cancel{r_2 X_0} + r_1 r_2 X_1 - r_1 r_2 Z_1 - \cancel{r_1 Z_0}}{r_2 - r_1} \\
&=& \frac{-r_1 X_0 - r_2 Z_0 + r_1 r_2 X_1 - r_1 r_2 Z_1}{r_2 - r_1}
\end{eqnarray}

Encontramos que la solucion es:

\begin{eqnarray*}
      X_t &=&   C_1 \left ( \frac{1}{r_1} \right )^t + C_2   \left ( \frac{1}{r_2} \right )^t
      \\  && \frac{1}{r_1 - r_2} 
      \left [ (r_1 - r_2) Z_t +  \left (r_1 \frac{1}{r_2} - r_2 \frac{1}{r_1} \right ) Z_{t-1} 
        \right .   \\
        && \left . +  \left (r_1 \frac{1}{r_2^2} - r_2 \frac{1}{r_1^2} \right ) Z_{t-2}  
        \cdots + 
        \left (r_1 \frac{1}{r_2^n} - r_2 \frac{1}{r_1^n} \right ) Z_{t-n} + \cdots 
      \right ] \quad  , \quad \text{with} \\
      C_1 &=& \frac{ - r_1 X_0 - r_2 Z_0 + r_1 r_2 X_1 - r_1 r_2 Z_1   } {r_2 - r_1}  \\
      C_2 &=& \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 Z_1 + r_1 Z_0 }{r_2 - r_1} \, .
    \end{eqnarray*}


Si fueramos a programar esto escribimos:

\begin{eqnarray}
      X_0 &=&   C_1 +  C_2 + Z_0  \\
      X_1 &=& \frac{C_1}{r_1} + \frac{C_2}{r_2} + Z_1 +  \frac{r_1 + r_2}{r_1 r_2} Z_0  \\
      &\vdots&  \\
      X_n &=& C_1 \left ( \frac{1}{r_1} \right ) ^n + C_2  \left ( \frac{1}{r_2} \right )^n
      + Z_n +  \nonumber \\
      &&  \frac{1}{r_1-r_2} \left [ \left ( r_1 \frac{1}{r_2} - r_2 \frac{1}{r_1}  \right ) 
        Z_{n-1} + \cdots +  \left ( r_1 \left ( \frac{1}{r_2} \right)^n - r_2 \left ( \frac{1}{r_1}  
      \right )^n \right ) Z_0 \right ]   \\  \\
      C_1 &=& \frac{-r_1 X_0 - r_2 Z_0 + r_1 r_2 X_1 - r_1 r_2 Z_1 }{r_2 - r_1}  \nonumber \\
      C_2 &=& \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 Z_1 + r_1 Z_0 }{r_2 - r_1}.  \\
       \\
    \end{eqnarray}


Es posible que las soluciones $r_1, r_2$ sean complejas y debemos lidiar con este problema. Esta solucion es perfectamente valida y se puede implementar en Python aun con $r_1, r_2$ complejas, pero vamos estudiar el caso $r_1, r_2$ complejas por separado por que ofrece simplificaciones ademas de una implementacion donde no hay que usar numeros complejos.
