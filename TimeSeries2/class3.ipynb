# Otras distribuciones importantes
## Distribucion Gamma $\Gamma$ 
La distribucion Gamma $\Gamma$ esta dada por

$$ p_{k, \theta} = \frac{1}{\Gamma(k) \theta^k} x^{k-1} \mathrm{e}^{-x/\theta}, $$

Donde:  
* [Gamma:](https://es.wikipedia.org/wiki/Funci%C3%B3n_gamma) la funccion $\Gamma$ es la funcion de Euler.

* $k>0$: se conoce como el **shape parameter** 
* $\theta>0$ se conoce como la escala **scale**.

El dominio de la distribucion $\Gamma$ esta en $(0,\infty)$

Ejemplo en Python

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1,2)
fig.suptitle("Gamma $\Gamma$ function")
fig.set_size_inches(14.5, 10.5)

# definimos varaias distribuciones
x = np.linspace(-0.1, 20, 100)

# a= k, scale = theta
y1 = stats.gamma.pdf(x, a=0.5, scale=2)
y2 = stats.gamma.pdf(x, a=1, scale=2)
y3 = stats.gamma.pdf(x, a=3, scale=2)
y4 = stats.gamma.pdf(x, a=6, scale=2)

axs[0].plot(x, y1, label="shape=0.5 scale=2", color='k')
axs[0].plot(x, y2, label="shape=1 scale=2", color='r')
axs[0].plot(x, y3, label="shape=3 scale=2", color='g')
axs[0].plot(x, y4, label="shape=6 scale=2", color='b')



axs[0].legend(loc="upper right")

y1 = stats.gamma.pdf(x, a=0.5, scale=1)
y2 = stats.gamma.pdf(x, a=1, scale=1)
y3 = stats.gamma.pdf(x, a=3, scale=1)
y4 = stats.gamma.pdf(x, a=6, scale=1)

axs[1].plot(x, y1, label="shape=0.5 scale=1", color='k')
axs[1].plot(x, y2, label="shape=1 scale=1", color='r')
axs[1].plot(x, y3, label="shape=3 scale=1", color='g')
axs[1].plot(x, y4, label="shape=6 scale=1", color='b')

axs[1].legend(loc="upper right")


# etiquetas
axs[0].set(xlabel="domain ($x$)", ylabel="Gamma $\Gamma(x)$ function")
axs[1].set(xlabel="domain ($x$)", ylabel="Gamma $\Gamma(x)$ function")

plt.show()
```

## Distribucion Chi-cuadrado $\chi^2$
Esta distribucion es un caso particular de la distribucion $\Gamma$ donde $\theta=2$, y $k \to k/2$. Es decir

$$ p_{k/2, \theta} = \frac{1}{\Gamma(k/2) 2^{k/2}} x^{k/2-1} \mathrm{e}^{-x/2}, $$ 
La funcion $\chi$ cuadrado, solo depende de un parametro que es $k$, *degrees of freedom** (grados de libertad).

image.png

La interprertacion es que a medida que se incrementa el numero $k$ (grados de libertad) la funcion se "estira" hacia la derecha y pierde altura. 

Por esta funcion es tan importante:

En la estadistica hay una tecnica que se llama **hypothesis testing**. Donde se prueba una hipotesis, se asume la hipotesis  [nula ](https://en.wikipedia.org/wiki/Null_hypothesis) $H0$ en contraste con otra hipotesis alternativa $H1$. Para decidir sobre esta hipotesis se determina una distribucion (muchas veces se usa la $\chi$ cuadrado. Existe un parametro llamado $p$ que generalmente es 0.05 (5%) el cual es usado como humbral para determinar la validez de la hipoteis. En curso discutiremos esto con mas detalle.

Existe una lista de propiedades que pueden estimar con la distribucion $\chi$ cuadrado.

* Se probar que ciertos datos tienen una PDF.
* Verificar si una poblacion tiene una media o varainza deterrereminada
* Se puede evaluar que tan bueno es un ajuste de unos datos a un modelo.
* Se puede verificar el grado de independencia de variables.
* Se puede verificar que una serie de tiempo es aleatoria o no. Es importante que los residuales no tengan correlacion. Esto lo usamos mas adelante en el curso.

# Cumulative Distribution Funcion: Funciones de distribucion acumuladas.

Una vez se tienen las densidades, estas se pueden utilizar para obtener probabilidades de eventos. Por definicion


$$ F_X(x) = P(X \le x)  $$

Puede ser discreta, o continual. Tambien podemos escribir

$$ P(X \le x ) = P(-\infty < X \le x)$$
Si fuese discreta tenemos:

$$F_X(x) = \sum_{x_k \le x} p(x_k)  \quad , \quad p(x_k)=P(X=x_k) $$
si es continua

$$ F_X(x) = \int_{-\infty}^x p(t) dt)  \quad , \quad p(x) = P(X=x) $$

Por el teorema fundamental del calculo

$$ p(x) = \frac{d F_X(x)}{dx}$$.

Podemos pensar el la probabilidad  en un intervalo

$$p( a \le X \le b) = F_{X}(b) - F_{X}(a) = \int_a^b p(x) dx $$

Es claro que

$$\lim_{x \to -\infty} F_X(x) = 0  \quad , \quad 
\lim_{x \to \infty} = 1 $$.
La funcion $F$ es monotona creciente.

Las funciones acumulativas no siempre tiene un forma  anlitica cerrada.

Por ejemplo, la funcion gaussiana (normal) no es integrablle analiticamente.

Es decir:

# Medidas de tendencia central en datos finitos.
En la practica los datos pueden ser jigantescos. Queremos interprertar los datos con unos pocos numeros (con el fin de simplicar). A los datos se les llama **poblacion**. Usted quiere hacer inferencias sobre la poblacion. A las inferencias sobre la poblacion se les llama **parametros**. Cuando las mediciones se hacen sobre una muestra, se llaman **estadisticas**. 

Vamos a ver ejemplos con muestras.

* **media** (mean). Es la medida central mas comun. Es el promedio de los valores de todas muestras. Es apropiadad para distribuciones simetricas.  Dado una muestra $X$ con valores $\{x_i\}$ reales, con $i \in I$. La media, que se nota con $\overline{x}$ se calcula como

$$ \overline{x} = \sum_{i \in I} \frac{x_i}{n} $$
donde $n = \# X$.

* **mediana** (mediana). No es promedio, es el centro de las muestras. Si tenemos una secuencia ordenada (de menor a mayor) 
$x_1, x_2, \cdots, x_m$. La muestrar de la mitad es $x_k$, $k=(m+1)/2$. Eso si $m$ es impar. Que pasa si $m$ es par? En este caso se usa el promedo de los dos centrales.

Es decir:

$$ M = \left \{ 
    \begin{array}{cc}
    x_{k+1} &  m = 2k + 1 \quad , \quad \text{odd} \\
    \frac12 ( x_k + x_{k+1}) & m = 2k  \quad , \quad \text{even} 
    \end{array}
\right .$$

La media es importante por que ignora los "outliers" (ruido jigante, **valor atipico**).

Por ejemplo tenemos 3 muestras $1, 10 , 10^6$ ($10^6$ es un outlier). El promedio esta dado por

$$ \overline{x} = \frac{1 + 10 + 10^6}{3} = 333337 $$
La media esta **sesgada* hacia valores muy grandes, que son atipicos.

La mediana es $10$, que es insensible a valores atipicos.

Si tenemos cuatro muestras $\{1,10,20, 10^6\}$. 
La media la calculamos como

$$ m = \frac{10 + 20 }{2} = 15 $$

Vemos que como la media no se deja **sesgar** se puede usar para filtrar datos con valorers **atipicos**. Existe un filtro que se llama  [median filter](https://en.wikipedia.org/wiki/Median_filter). La idea es crear una ventana corrediza (running window) con un tamano $k$ y de cada paso se toma la mediana.

Ejemplo $\{1, 5, 10, 10^6, -3 ,3 , 1 \}$.
Tomemos una ventana con $k=3$.

* $(1,5,10) \to 5$
* $(5,10,10^6) \to 10$
* $10, 10^6, -3 \to 10$
* $10^6 , -3, 3 \to 3$
* $-3, 3, 1 \to 1$

Las muestras seleccionadas son $5,10,10,3,1$. Son todas correspondientes a los datos pero se filtro el $10^6$. 

Si se va a escribir un algoritmo de **median filter**

* se toman los datos $X$, asumiendo "fold =k".
* Se crea una ventana corrediza de tamano $k$
* En esta ventana se toma el centro. 

Vimos dos medidas centrales importantes. Hay mas que no vamos a discutir. Por ejemplo **quantiles** (el 20% mejor, el 40% mejor, el 60% mejor, el 80% mejor y el mejor (100 % mejor). De los **quantiles** el comun es los **quartiles** (25%, 50%, 75%, 100%). El %50 es la media. El 25% es la media de los datos entre el mas pequeno y la media, el 75% es la media entre la media y el mas grande.

## Medidas de dispersion. Que tan regados estan los datos. 

Ejemplos:

* **rango**: Es el maximo menos el minimo. Ej: $\{1,-10,20,30\}$ el rango es $30-(-10)=40$. 

* **varianza**: Se define con la formula

$$s^2 = \frac{1}{n-1} \sum (x_i - \overline{x})$$ 
Pareciera que tuvieramos que dividir por $n$ y no por $n-1$ para tomar una especia de promedio. Pero no es
asi para que sea un estimador no sesgado.
[unbiased estimator](https://en.wikipedia.org/wiki/Bias_of_an_estimator)

* Desviacion estandard $s=\sqrt{s^2}$.

$$s = \sqrt{\frac{1}{n-1} \sum (x_i - \overline{x})}$$ 

La media y varianza estan relacionadas a dos conceptos
importantes.

* **valor esperado** (esperanza) , **expected value**.

$$ \mathbb{E}[X] = \sum_{i \in I} x_i p_i$$
si la probabilidad es uniforme esto es la media. Esto para variables discretas.

Para continuas

$$ \mu = \mathbb{E}(X) = \int_{-\infty}^{\infty} x p(x) dx$$

Hay dos propiedades

*  $\mathbb{E}(X+Y) = \mathbb{E}(X) + \mathbb{E}(Y)$
* $\mathbb{E}(\alpha X) = \alpha \mathbb{E}(X) \quad \alpha \in \mathbb{R}$. 

* La varianza

$$ \sigma_X = \mathbb{V}(X) = \mathbb{E}[(X-\mu^2)]$$

Se puede probar que

$$\mathbb{E}[ (X - \mu)^2] = \mathbb{E}[X^2] - \mu^2 $$

Si es discreta

$$ \sigma^2 = \mathbb{V}[X] = \sum_{i\in I} (x_i - \mu)^2 p_i$$

Si es continua

$$\sigma^2 = \int_{-\infty}^{\infty} (x - \mu)^2 p(x) dx $$

y

$$ \mathbb{V}(\alpha x) = \alpha^2 \mathbb{V}(X)$$.
Y la desviacion estandard es $\sigma$.
