## Empalme con la clase anterior

* **Varianza**:
De $X_t = \phi_1 X_{t-1} + \phi_2 X_{t-2} + Z_t$, $c=0$.

"A naive approach", usando propiedades de la varianza y la hipotesis de estacionarieidad debil

$$ \text{Var}{X_t}= \sigma^2$$.

$$ \sigma^2 = \phi_1^2 \sigma^2 + \phi_2^2 \sigma^2 + \sigma_Z + 2 \text{Cov}[]....  $$
Tarea: continuar este ejercicio hasta obtener $\sigma^2$.

Hagamos el problema por otro lado.

\begin{eqnarray}
\gamma_k &=& \mathbb{E}[X_t X_{t-k}] = \phi_1 \mathbb{E}[X_{t-1} X_{t-k}] + \phi_2 \mathbb{E}[X_{t-2} X_{t-k}] + \mathbb{E}[Z_t X_{t-k}] \\
&=& \phi_1 \gamma_{k-1} + \phi_2 \gamma_{k-2} + \mathbb{E}[Z_t X_{t-k}] 
\end{eqnarray}

donde

\begin{eqnarray}
\mathbb{E}[Z_t X_{t-k}]  = \left \{
    \begin{array}{cc}
    \sigma_Z^2 & \text{si  } k = 0  \\
    0 & \text{si } k=1,2, ,\cdots
    \end{array}
    \right .
\end{eqnarray}

Recuerden que $\gamma_t = \gamma_{-t}$ (simetrica).


Combinemos estas dos ultimas ecuaciones.

\begin{eqnarray}
\sigma^2 &=& \gamma_0 = \phi_1 \gamma_1 + \phi_2 \gamma_2 + \sigma_Z^2  \quad , \quad \text{recuerde que } k=0 \quad , \quad (1)  \\
\gamma_k &=& \phi_1 \gamma_{k-1} + \phi_2 \gamma_{k-2} \quad , \quad \text{para } k=1,2, 
\end{eqnarray}
 
 Para $k=1$,

 $$\gamma_1 = \phi_1 \gamma_0 + \phi_2 \gamma_1 $$

 despejemos $\gamma_1$,

 $$ \gamma_1 =  \frac{\phi_1 \gamma_0}{1-\phi_2} =\frac{\phi_1 \sigma^2}{1-\phi_2} $$

 Para $k=2$.

 $$\gamma_2 =  \phi_1 \gamma_1 + \phi_2 \gamma_0 = \phi_1 \frac{\phi_1 \sigma^2}{1- \phi_2 } + \phi_2 \sigma^2 = \sigma^2 \left (   \frac{\phi_1^2}{1 - \phi_2}  + \phi_2 \right ) = \sigma^2 \left ( \frac{\phi_1^2 + \phi_2(1- \phi_2)}{1 - \phi_2}\right ) $$
 
Reescribiendo la (1) con este $\gamma_2$. Reemplazando  $\gamma_1$ y $\gamma_2$ de las ecuaciones de

$$\sigma^2 = \gamma_0 = \phi_1 \frac{\phi_1 \sigma^2}{1-\phi_2}  +\sigma^2 \left ( \frac{\phi_1^2 + \phi_2(1- \phi_2)}{1 - \phi_2}\right ) \phi_2  + \sigma_Z^2 $$

En este momento todo esta en funcion de $\sigma^2, \sigma_Z^2$ y de los coeficientes $\phi_1, \phi_2$ entonces podemos despejar $\sigma^2$ (asumimos que conocemos $\sigma_Z^2$. 




Para despejar movemos a la izquierda todos los terminos con $\sigma^2$ y lo factorizamos y obetenemos

$$\sigma^2 \left [ 1 - \frac{\phi_1^2}{1 - \phi_2} - \phi_2 \left ( \frac{\phi_1^2 + \phi_2 (1 - \phi_2}{1 - \phi_2}  \right ) 
    \right ] = \sigma_Z^2 $$

Antes de despejar $\sigma^2$ simplifiquemos la expresion:

\begin{eqnarray}
1 - \frac{\phi_1^2}{1 - \phi_2} - \phi_2 \left ( \frac{\phi_1^2 + \phi_2 (1 - \phi_2)}{1 - \phi_2}  \right ) &=& \frac{1 - \phi_2 - \phi_1^2 - \phi_2 \phi_1^2 - \phi_2^2  + \phi_2^3}{1 - \phi_2}  \\
&=& \frac{(1 - \phi_2^2)  - \phi_1^2(1 + \phi_2) + \phi_2(\phi_2^2 -1) }{1 - \phi_2} \\
&=& \frac{(1 - \phi_2)(1+ \phi_2) - \phi_1^2 (1 + \phi_2) - \phi_2(1 - \phi_2)(1+\phi_2)}{1 - \phi_2} \\
&=& \frac{(1 + \phi_2)(1 - \phi_2 - \phi_1^2 - \phi_2(1 - \phi_2))}{1 - \phi_2} \\
&=& \frac{(1 + \phi_2) ( 1 - 2 \phi_2 + \phi_2^2 - \phi_1^2}{1 - \phi_2} \\
&=& \frac{(1 + \phi_2)((1-\phi_2)^2 - \phi_1^2)}{1 - \phi_2} \\
&=& \frac{(1 + \phi_2)(1 - \phi_2 - \phi_1)(1 - \phi_2 + \phi_1)}{1 - \phi_2}
\end{eqnarray}

$$\sigma_2 = \frac{(1 - \phi_2) \sigma_Z^2}{(1 + \phi_2)(1 - \phi_2 - \phi_1)(1 - \phi_2 + \phi_1)} $$

Hay restricciones:

$$ \phi_1 + \phi_2 < 1  \quad , \quad \phi_2-\phi_1 < 1 \quad , \quad |\phi_1|<1 $$

* **Funcion de autocorrelacion ACF** :

Arrancamos de

$$ \gamma_k = \phi_1 \gamma_{k-1} + \phi_2 \gamma_{k-2} + \mathbb{E}[Z_t X_{t-k}] $$
Donde


\begin{eqnarray}
\gamma_0 &=& \sigma_0^2  \quad , k=0 \\
\gamma_k &=& \phi_1 \gamma_{k-1} + \phi_2 \gamma_{k-2} \quad , \quad k=1,2, \cdots 
\end{eqnarray}

Este problema no es nuevo. Por es una ecuacion en diferencias (que ya la resolvimos). Excepto por que la incongnita ahora es $\gamma_k$ en vez de $X_k$.
Si usamos el operador backshift, tenemos

$$ (1 - \phi_1 B - \phi_2 B^2) \gamma_k = 0 $$
La solucion esta en la clase 13, ecuacion (*****). Primero halle
$r_1, r_2$ soluciones de la ecuacion caracteristica.

Asi quedan las cosas:

\begin{eqnarray}
\gamma_0 &=& \sigma^2 = \frac{(1 - \phi_2) \sigma_Z^2}{(1 + \phi_2)(1 - \phi_2 - \phi_1)(1 - \phi_2 + \phi_1)} \\
\gamma_1 &=& \frac{\phi_1 \sigma^2}{1 - \phi_2} \\
\gamma_2 &=& \sigma^2 \left ( \frac{\phi_1^2 + \phi_2(1 - \phi_2)}{1 - \phi_2} \right )
\gamma_k &=& C_1 \left ( \frac{1}{r_1} \right )^t + C_2 \left ( \frac{1}{r_2} \right )^t \\
C_1 &=&  r_1 \frac{\sigma_2 - r_2 \gamma_1}{r_1 - r_2}  \quad , \quad C_2 = \frac{r_2 ( r_1 \gamma_1 - \sigma_2 )}{r_1 - r_2}  
\end{eqnarray}

### Ejemplo 1. Un modelo **inestable**
Asumamos $c=0$, $\phi_1=1, \phi_2=0.5$. Observe que este modelo no es inestable por que no califica a las ecuaciones de estabilidad. 
$\phi_1 + \phi_2 = 1.5$ y $1.5$ no es menor que uno.

El modelo es

$$X_t = X_{t-1} + 0.5 X_{t-2} + Z_t $$


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1) # para ser consistentes cada vez
mu, sigma = 0, 1
N=1000
Z = np.random.normal(mu, sigma, N)

phi_1 = 1.0
phi_2 = 0.5

X= np.zeros(N)


# condiciones iniciales
X[0] = Z[0]
X[1] = Z[1]

for t in range(2, N):
    X[t] = Z[t] + phi_1*X[t-1] + phi_2 * X[t-2]
```

```python
# etiquetas
plt.xlabel("numero de muestra")
plt.ylabel("valor de la muestra")
plt.title("AR(2), con $\phi_1$ =1, $\phi_2$=0.5")
plt.plot(X);
plt.show();
```

### Ejemplo 2.  Modelo estable con dos raices reales distintas.
$\phi_1 = 0.7$, $\phi_2=0.2$

$$ \phi_1 + \phi_2 < 1  \quad , \quad \phi_2 - \phi_1 < 1 \quad, \quad |\phi_2| < 1 $$.

Veamos

- [x] $\phi_1 + \phi_2 = 0.9 < 1 $ 
- [x] $\phi_2 - \phi_1 = -0.5 < 1$
- [x]  $|\phi_2| = 0.2 < 1 $

Como se resuelve este problema:

* Resolver el problema homogeno
* Encontrar una solucion particular
* Encontrar la solucion como la suma de las dos anteriores aplicando las condiciones iniciales.

Las raices del polinomio caracteristico

$$ p(B)= (1 - 0.7 B - 0.2 B^2) = 0 $$

Veamos el codigo en python.

$$ r_{\pm} = \frac{0.7 \pm \sqrt{0.7^2 + 0.8}}{-0.4}$$
$$r_1 = -4.589 \quad , \quad r_2 = 1.089 $$

Por ahora concentremonos en el problema homogeneo

\begin{eqnarray}
X_{th} &=& C_1 \left ( \frac{1}{r_1} \right )^t + C_2 \left ( \frac{1}{r_2}  \right )^t \\
C_1 &=& \frac{r_1 ( X_0 - r_2 X_1)}{r_1 -r_2}  \quad , \quad C_2 = \frac{r_2( r_1 X_1 - X_0 )}{r_1 -r_2}
\end{eqnarray}

```python
# encontremos la ecuacion recursiva
X = np.zeros(N)
X[0]= 1.62  # tomado de Z[0] en el codigo anterior
X[1] = -0.611

phi_1 = 0.7
phi_2 = 0.2

for t in range(2, N):
    X[t] = phi_1 * X[t-1] + phi_2 * X[t-2]

```

```python
# etiquetas
plt.xlabel("numero de muestra")
plt.ylabel("valor de la muestra")
plt.title("AR(2), con $\phi_1$ =1, $\phi_2$=0.5")
plt.plot(X);
plt.show();
```

```python
# solucion analitica
XH = np.zeros(N)
# raices del polinomio caracteristico
r1 = (phi_1 + np.sqrt(phi_1**2 + 4*phi_2))/(-2*phi_2)
r2 = (phi_1 - np.sqrt(phi_1**2 + 4*phi_2))/(-2*phi_2)

# r1 = (phi_1 + np.sqrt( phi_1**2 + 4*phi_2)))/(-2*phi_2)

print("r1 = ", r1)
print("r2 = ", r2)
coeff = 1.0/(r1 - r2)

# condiciones iniciales
XH[0] = X[0]
XH[1] = X[1]

# coeficientes
C1 = coeff*( r1*( XH[0] - r2*XH[1]))
C2 = coeff*( r2*( r1*XH[1] - XH[0]))

# loop
for t in range(2,N):
    XH[t] = C1*(1/r1)**t + C2*(1/r2)**t


```

```python
print(phi_1,phi_2)
```

```python
# etiquetas
plt.xlabel("numero de muestra")
plt.ylabel("valor de la muestra")
plt.title("AR(2), con $\phi_1$ =1, $\phi_2$=0.5")
plt.plot(XH);
plt.show();
```

```python
# etiquetas
plt.xlabel("numero de muestra")
plt.ylabel("valor de la muestra")
plt.title("Error, iterativo vs analitico AR(2), con $\phi_1$ =1, $\phi_2$=0.5")
plt.plot(X-XH);
plt.show();
```
