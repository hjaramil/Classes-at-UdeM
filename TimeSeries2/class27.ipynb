# Modelo ARMA(1,1)
$$X_t = Z_t + \phi_1 X_{t-1} + \beta_1 Z_{t-1}. $$

Recuerden que debemos calcular

$$X_t X_{t-\tau} =  Z_t X_{t-\tau} + \phi_1 X_{t-1} X_{t - \tau} + \beta_1 Z_{t-1} X_{t - \tau}. \quad (0) $$

Estudiamos esta expresio termino por termino de la derecha.

El primer termino es y tomando la esperanza

$$ \mathbb{E}[Z_t X_{t-\tau} ] = \mathbb{E}[Z_t Z_{t-\tau} ]+ \phi_1 \mathbb{E}[ Z_t X_{t-\tau-1} ] + \beta_1 \mathbb{E}[Z_t Z_{t-\tau-1} ] .  $$

Vamos por partes

* $\tau=0$:

$$ \mathbb{E}[Z_t X_{t} ] = \mathbb{E}[Z_t Z_{t} ]+ \phi_1 \mathbb{E}[ Z_t X_{t-1} ] + \beta_1 \mathbb{E}[Z_t Z_{t-1} ] = \sigma_Z^2  + 0 + 0 .  \quad (1) $$

* $\tau \ge 1 $:

$$ \mathbb{E}[Z_t X_{t-\tau} ] = \mathbb{E}[Z_t Z_{t-\tau} ]+ \phi_1 \mathbb{E}[ Z_t X_{t-\tau-1} ] + \beta_1 \mathbb{E}[Z_t Z_{t-\tau-1} ]  = 0  +0 + 0 \quad (2 ) $$

Ahora vamos para el ultimo termino $\mathbb{E}[Z_{t-1} X_{t-\tau}]$. Tambien lo consideramos por partes

* $\tau=0$: 

\begin{eqnarray}
 \mathbb{E}[Z_{t-1} X_t] &=& \mathbb{E}[Z_{t-1} ( Z_t + \phi_1 X_{t-1} + \beta_1 Z_{t-1})] \\
 &=& \mathbb{E}[Z_{t-1} Z_t] + \phi_1 \mathbb{E}[Z_{t-1} X_{t-1}] + \beta_1 \mathbb{E}[Z_{t-1} Z_{t-1}] \\
 &=& 0 + \phi_1 \sigma_Z^2 + \beta_1 \sigma_Z^2 \\
 &=& (\phi_1 + \beta_1) \sigma_Z^2 \quad (3)
\end{eqnarray}

La ecuacion (0) luego de tomar la esperanza es:

$$\gamma_{\tau}= \mathbb{E}[X_t X_{t-\tau}] =  \mathbb{E}[Z_t X_{t-\tau}] + \mathbb{E}[\phi_1 X_{t-1} X_{t - \tau}] + \beta_1 \mathbb{E}[Z_{t-1} X_{t - \tau}]. \quad  (4)  $$

$$\gamma_{0}= \mathbb{E}[X_t X_{t}] =  \mathbb{E}[Z_t X_{t}] + \mathbb{E}[\phi_1 X_{t-1} X_{t}] + \beta_1 \mathbb{E}[Z_{t-1} X_{t}]. \quad  (5)  $$

Usamos la ecuacion (1) la ecuacion y la (1) y la (3) en 

$$\gamma_0 =  \sigma_Z^2 + \phi_1 \gamma_1  + \beta_1   (\phi_1 + \beta_1) \sigma_Z^2 \quad (6) $$

Ahora bien si

* $\tau = 1$: 
Debemos evaluar el ultimo termino de la ecuacion (4).






$$\mathbb{E}[Z_{t-1} X_{t-1}] = \sigma_Z^2 $$
La ecuacion (4) queda

$$\gamma_{1}= \mathbb{E}[X_t X_1] =  \mathbb{E}[Z_t X_{t-1}] + \mathbb{E}[\phi_1 X_{t-1} X_{t - 1}] + \beta_1 \mathbb{E}[Z_{t-1} X_{t - 1}] =  0 + \phi_1 \gamma_0 + \beta_1 \sigma_Z^2 \quad (7) $$

Reemplazamos 7 en 6.

$$ \gamma_0 = \sigma_Z^2 + \phi_1 ( \phi_1 \gamma_0 + \beta_1 \sigma_Z^2) + \beta_1( \phi_1 +  \beta_1) \sigma_Z^2 = \gamma_0 \phi_1^2 + \sigma_Z^2( 1 + \phi_1 \beta_1 + \beta_1^2 + \phi_1 \beta_1) $$

Todo esto esta en terminos de $\sigma_0$, el cual toca despejar:

$$\gamma_0 = \frac{1 + 2 \phi_1 \beta_1 + \beta_1^2}{1 - \phi_1^2 }  \sigma_Z^2 $$

Para $\gamma_1$ vamos a la ecuacion (7).

\begin{eqnarray}
 \gamma_1 &=& \left [  \phi_1 \frac{1 + 2 \phi_1 \beta_1 + \beta_1^2}{1 - \phi_1^2 } + \beta_1  \right ] \sigma_Z^2 \\
 &=& \frac{\phi_1 + 2 \phi_1^2 \beta_1 + \phi_1 \beta_1^2 + \beta_1 - \beta_1 \phi_1^2}{1 - \phi_1^2 } \sigma_Z^2 \\
  &=& \frac{\phi_1 + \phi_1^2 \beta_1 + \phi_1 \beta_1^2 + \beta_1 }{1 - \phi_1^2 } \sigma_Z^2 \\
 &=& \frac{\phi_1 ( 1 + \phi_1 \beta_1 ) + \beta_1(1 + \phi_1 \beta_1)}{1 - \phi_1^2 }  \sigma_Z^2 \\
 &=& \frac{(\phi_1 + \beta_1)(1 + \phi_1 \beta_1)}{1 - \phi_1^2} \sigma_Z^2
 \end{eqnarray}


 Que pasa si $\tau>1$?

 De la clase 25:

 $$ \gamma_{\tau} = \sum_{i=1}^p \phi_i \gamma_{\tau - i}  \quad , \quad \text{simpre que   } \tau \ge q+1. $$

 Como p=1,

 $$\gamma_{\tau} = \phi_1 \gamma_{\tau-1}. $$

 En conclusion el ACF del ARMA(1,1) esta dado por (normalizando por $\gamma_0=\sigma^2$)

 \begin{eqnarray}
\rho_{\tau} =
\left \{
    \begin{array}{cc}

    \end{array}.
    \right .
 \end{eqnarray}

## SSE: Suma de los errores cuadrados.
Este, de alguna forma, es como un "least squares" que nos lleva a las ecuaciones de Yule-Walker (recuerden la tarea de Norman Wiener).
Vamos a ver que los dos no son una copia del mismo.

Veamos una ilustracion con un modelo AR(p) para evitar complicaciones.

$$X_t = Z_t + \phi_1 X_{t-1} + \phi_2 X_{t-2} + \cdots +  \phi_p X_{t-p}.$$$$ $$

ESte es un estimado y como tal lo podemos llamar $\hat{X_t}$ y los datos reales
los llamamos $X_t$, el error en suma de cuadados es:

$$ \text{SSE}(p) =\| X_t - \hat{X_t} \|^2 = \sum_{i=1}^N (X_i - \hat{X_{i}})^2.$$

Observacion de notacion: El indice "$t$" a la izquierda indica que $X_t$ es una serie de tiempo (un vector) a la derecha el indice "$i$" indica que $X_i$ es una muestra, la muestra $i$ (componente $i$) del vector $X_t$.


El problema que queremos resolver es hallar p Tal qeu 


$$ \min_p \text{SSE}(p) =\min_p \| X_t - \hat{X_t} \|^2 = \min_p \sum_{i=1}^N (X_i - \hat{X_{i}})^2.$$

Veamos esto en la practica con un modelo AR(3) o  ARMA(3,0)

````python
# importar librerias
import statsmodels as sm
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.regression.linear_model import yule_walker
from statsmodels.tsa.stattools import pacf
# from statsmodels.regression.lineal_model import yule_walker

import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl
mpl.rc('lines', linewidth=0.5)

np.random.seed(1)  # no estoy seguro si se usa o no. Yo creo que no se usa

phi_1 = 0.9
phi_2=-0.6
phi_3=0.3


N=1000 # numero de muestras

arparms = np.array([phi_1, phi_2, phi_3])
maparms = np.array([0])

ar = np.r_[1, -arparms]
ma = np.r_[1, -maparms]

arma_process = ArmaProcess(ar, ma)
print("es el modelo estacionario?", arma_process.isstationary)
print("es el modelo estabel?", arma_process.isinvertible)  # estable
print("raices del polinomio caracteristico de AR(3)", arma_process.arroots ) # raies del polinomio caracteristico de AR(3)
y = arma_process.generate_sample(N, scale=1)  # datos, 


````

````python
# definicion de la funcion SSE:

def SSE(data, k, seed, N):  # la semilla es necesaria?
    # data son los datos
    # k es el numero iteraciones para estimar p, p<=k
    # seed: para reproducir los datos
    # N: numero de muestras

    # hay un ciclo de posibilidades
    # inicializacion del vector

    ssevec = np.zeros(k+1) # es mas phi_0, phi_0, phi_1 , phi_0 corresponde a c

    for i in range(1, k+1):
        np.random.seed(seed)

        # obtener los coeficientes con YW
        phi, sigma = yule_walker(data, i, method='mle')
        arparms = phi
        maparms = np.array([0])
        ar = np.r_[1, - arparms]
        ma = np.r_[1, -maparms]
        arma_process = ArmaProcess(ar, ma)
        y = arma_process.generate_sample(N, scale=1)

        sse = np.sum( (y-data)**2)
        ssevec[i]=sse

    return ssevec[1:]



````

````python
# probemos la funcion
k=6
N=1000
seed=1
myssevec=SSE(y,k,seed,N)
myssevec
````

````python
min_value=min(myssevec)

min_index = list(myssevec).index(min_value)
print(f"el p ideal es {min_index+1} con SSE igual a {min_value}")
````
