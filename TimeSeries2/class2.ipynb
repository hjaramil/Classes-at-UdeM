La **probability distribution** es una funcion que mapea todos los posibles eventos a una probabilidad. Hay dos tipos de distribuciones de probabilidad: **discreta** y la **continua**. La discreta es para un numero finito de eventos, la continua es en intervalos. A la discreta se le conoce como 
**Proability Mass Function*** (PMF) y a la continua como **Probability Density Function** (PDF). No confundir con Portable Data Format (de los archivos PDF).

## Distribuciones de probabilidad discretas. 
Recuerde que la probabilidad del universo es 1. Es decir:

$$ P(S) = \sum_{i \in I}  P(X=x_i) = \sum_{i \in I} p_i = 1 $$

Probabilidad de un evento. Sea $T \subset S$. Entonces

$$ P(T) = \sum_{i \in I_T} p_i$$

Como es discreta uno puede calcular la probabilidad como la suma de las probabilidades de los miembros validos. 

Ejemplo: Asuma que un dado tiene las siguientes salidas 1,2,3,4,5,6. Si el dado no esta sesgado, la probabilidad de cada evento es $$P(X=x_i) = p_i = 1/6 $$.
Entonces la probabilidad de que los eventos sean $2 \le X \le 5$ es

$$P(2 \le X \le 5) = P(X=2 \;o\; X=3 \;o\; X=4 \;o\; X=5)=\sum_{i=2}^5 p_i = 4 \frac16 = \frac{2}{3} $$

### Ejemplos de distribuciones discretas.

* La distribucion de probabilidad uniforme. Aca todos los eventos tienen las mismas posibilidades. La distribucion de probabilidad es constante  $p$. 
Como la probabilidad del universo es 1, si tenemos $m$ eventos tenemos que

$\sum_{i=1}^m p_i = 1 = m p$ de donde

$$ p = P(X=x) = \frac1m $$


En el caso de el dado $m=6$.

* La **binomial**.  Esta distribucion esta atribuida a fenomenos que son binarios. Solo toman dos valores. Por ejemplo una moneda. A estos experimentos se les llama **Bernoulli trials**.   Aca esta la descripcion de este fenomeno.

    * Hay un numero fijo de intentos $n$
    * Los intentos $n$ son independientes
    * Solo hay dos salidas (binario)
    * Las probabilidades de cada evento no cambian entre intentos (trials). Por ejemplo si la probabilidad de que ocurra un evento es $p$, esta siguie siendo $p$ para cualquiera de los otros y $1-p$ para que no ocurra ese evento.

Con esta informacion tenemos

$$ P(X=k) = p(k) = p_k = \binom{n}{k} p^k q^{n-k} = \binom{n}{k} p^k (1-p)^{n-k}. $$

Donde

$$ \binom{n}{k} = \frac{n!}{(n-k)! k!} $$.

Ejemplo en Python

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# fijar valores de n y p
n=10 # numero de intentos
p=0.8  # moneda sesgada

k_values=list(range(n+1))
dist = [binom.pmf(k,n,p) for k in k_values]

# grafiquemos la distribucion binomial
plt.title("binomial con $p=0.8$")
plt.xlabel("k")
plt.ylabel("P(X=k")
plt.bar(k_values, dist, color='blue')
plt.show()
```

## Distribuciones de probabilidad continuas (PDF)

Similarmente a como hicimos en las discretas, pero cambiamos sumas por integrales.

$$\sum  \int$$

Tenemos todos los eventos ocurren en el continuo $[a,b]$

$$ \int_a^b p(x) dx = 1  $$ o mejor aun

$$\int_{-\infty}^{\infty} p(x) dx = 1  $$ donde aqueyos eventos que no existan tiene $p(x)=0$. 

Si quiere calcular la probabilidad en un intervalo $[u,v]$ esta se calcula con la formula

$$ p(u \le x \le v ) = \int_u^v p(x) dx  $$


Ejemplos:

* Distribucion de Probabilidad **uniforme**. En este caso todos los eventos tienen igual probabilidad. Asi que

$$\int_S p(x) dx = 1 $$
como $p=p(x)$ es constante

$$ \int_S p(x) dx = p \mu(S)$$ donde $\mu$ es la **medida**.
Por ejemplo si $S=[a,b]$ entonces $\mu(S) = b-a$.

Si $S=[a,b]$ tenemos que

$$ p(x) = p = \frac{1}{\mu(S)} =  \frac{1}{b-a} $$

* Distribucion normal. Es la mas importante en la estadistica parametrica. 

$$p(x; \mu, \sigma) = \mathcal{N}(\mu, \sigma^2) = \frac{1}{\sigma \sqrt{2 \pi}} \mathrm{e}^{-\frac12 \left ( \frac{x - \mu}{\sigma} \right )} $$

Aca $\mu$ se conoce como la **media** y $\sigma$ como la **desviacion estandard**. Por que esta distribucion.



image.png

En la grafica se observan dos distribuciones normales. La una con media 0 y la otrao con media 1 (en la media esta el pico). No se ve claro por que pero las desviaciones estandard son $0.5$ y $1$ respectivamente. Entre mas grande la desviacion estandar mas aplanada es la curva. 

De acuerdo a Walker (1985) y a Stahl (2006) dicen que fue De Moivre quien descrubrio la districion normal y no Gauss y que Laplace la conocia antes que Gauss. Los mayores atributos son a Gauss, de hecho muchos la conocen como la **campana Gaussiana**.


**Por que es tan importante la distribucion normal?**:

* Muchas de las distribuciones en la naturaleza son de este tipo: Por ejemplo estatura de los humanos, pesos de nacimiento de los humanos, satisfaccion en el trabajo, habilidad para leer, indice de inteligencia IQ. 

* Los metodos parametricos de la estadistica mas poderosos se basan en distribuciones normales. 

* El teorema del limite central CLT: Central Limit Theorem dice que:
Dado un conjunto de variables aleatorias $X_1, X_2, \cdots , X_n$ estadisticamente independientes, con media $\mu$ y varianza $\sigma^2$.
Si $\bar{X_n}$ ess la meda de la muestra $X_n$ entonces

$$Z = \lim_{n \to \infty} \sqrt{n} \left ( \frac{\bar{X_n} - \mu}{\sigma} \right ) $$

Entonces $Z$ es una distribucion **normal** con media $\mu$ y varianza $\sigma^2=1$. Si no se normaliza entoncese la media seria $\mu$ y la varianza $\sigma^2$. 

Este teorema realza la importancia de la distribucion normal. 

Ejemplo (Python): Pienses en $X_n$ como distribuciones binomiales con $n$ muestras.
La media y la desviacion estandar de estas esta dada por

$$\mu = n p  \quad , \quad \sigma=\sqrt{np(1-p)} $$

```python
import numpy as np
from scipy.stats import binom
from scipy.stats import norm
import matplotlib.pyplot as plt



# fijar valores
n=100
p=0.8

# valores de k
k_values = list(range(n+1))
# lista de pmf
dist = [binom.pmf(k,n,p) for k in k_values]

mean = n*p
std = np.sqrt(n*p*(1-p))

distribution = norm(mean, std)
probabilities = [distribution.pdf(v) for v in k_values]

# plot
plt.bar(k_values, dist, color='blue', label='discrete binomial')
plt.title('binomial con 100 experimentos de Bernoulli')
plt.plot(k_values, probabilities, color='red', label='Normal distribution')
legend = plt.legend(loc='upperre right', fontsize='x-small')
plt.show()
```

* Otro punto a favor de la distribucion normal es el de **minimos cuadrados**.
Tambien controversia, fue Gauss o Lagrange el que invento el metodo de minimos cuadrados? (Stigler, 1981).

En 1808 Gauss formulo el problema de minimos cuadrados de eforma estadistica.

Piense en una regresion lineal $A \theta =b$, donde $A$ ese una matriz $m \times n$, dodnde $\theta$ es un vector en $\mathbb{R}^n$ y $b \in \mathbb{R}^m$.
Tenemos que los columnas de $A$ son variables aleatorias $x_i$, el error se puede escribir como

$$ \epsilon_i = x_i^T \Theta - b $$. Gauss dijo que estos errores tenian una distribucion de probabilidad $\phi$ y que luego esta probabilidad era normal. Que los erroress eran estadisticamente independientes.

$$ P=\phi(\epsilon_1) \phi(\epsilon_2) \cdots \phi(\epsilon_n) =\prod_{i=1} \phi(\epsilon_i)$$

Cada una de ellas se comporta como

$$ \phi(\epsilon_i) = \frac{1}{\sqrt{2 \pi \sigma^2}}  \mathrm{e}^{-\frac{\epsilon_i^2}{2 \sigma^2}}$$

Al hacere la productoria encontramos

$$P = \prod_{i=1}^n \phi(\epsilon_i) = \frac{1}{\sqrt{(2 \pi)^n \sigma^{2n}}} \mathrm{e}^{-\frac{1}{\sigma^2} \sum_{i=1}^m (x_i \theta - b_i)^2} =\frac{1}{\sqrt{(2 \pi)^n \sigma^{2n}}} \mathrm{e}^{-\frac{1}{\sigma^2} \| A \theta - b \|^2} $$

Si quiero maximizar la probabilidad, me toca minimzar el exponente.
Es decir debo resolver el problema

$$ \min_{\theta} \| A \theta - b \|^2$$
