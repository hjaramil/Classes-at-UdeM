# Cross-covariance: covarianza cruzada, y autocovarianza

## Covarianza cruzada. De ahora en adelante, nuestras variables aleatorias son series de tiempo. Asumamos dos series de tiempo
$X_t$, $Y_t$. Si tomamos la covarianza entre ellas, como funcion del tiempo, esta se entiendo como **covarianza cruzada** (**cross-covariance**).
Es decir,

\begin{eqnarray}
 K_{XY}(t_1, t_2) &=& \text{Cov}(X_{t_1}, Y_{t_2}) \\
 &=& \mathbb{E}[ (X_{t_1} - \mu_{X_{t_1}}) (Y_{t_2} - \mu_{Y_{t_2}})] \\
 &=& \mathbb{E}[X_{t_1}, Y_{t_2}] - \mu_{X_{t_1}} \mu_{Y_{t_2}}.
\end{eqnarray}

Si $X_t$ y $Y_t$ son estacionarios entonces

\begin{eqnarray}
 K_{XY}(t_1, t_2) &=& 
 &=& \mathbb{E}[X_{t_1}, Y_{t_2}] - \mu_{X} \mu_{Y}.
\end{eqnarray}

Vamos a pensar en el "shift" (desplazamiento). Llamos $t = t_1$, $t_2= t_1 + \tau = t_1 + \Delta t$, donde $\tau=t_2 -t_1 = \Delta t$. 
Ahora vamos a pensar en terminos de  $t, \tau$.

Re-escribimos la ecuacion anterior como:
\begin{eqnarray}
 K_{XY}(t, \tau) &=& 
 &=& \mathbb{E}[X_{t}, Y_{\tau}] - \mu_{X} \mu_{Y}.
\end{eqnarray}

Si $X=Y$ entonces el termino es **autocovarianza**. Aca:


\begin{eqnarray}
 K_{XX}(t, \tau) &=& 
 &=& \mathbb{E}[X_{t}, X_{\tau}] - \mu_{X}^2 .
\end{eqnarray}

Formalicemos  el cocepto de estacionarieidad debil. Una serie de tiempo 
es estacionaria debilmente si:

* El primer momento, la esperanza se conserva con el tiempo. $\mathbb{E}[X_t] = \mathbb{E}[X_{t+\tau}]$. $\tau \in \mathbb{R}$ si el proceso es continuo o $\tau \in  \mathbb{Z}$ si el proceso es discreto (este curso).

* El segundo momento, la varianza se conserva. Es decir  $\mathbb{E}[(X_t- \mu)^2] = \mathbb{E}[X_{t+\tau} - \mu^2 ]$. $\tau \in \mathbb{R}$ si el proceso es continuo o $\tau \in  \mathbb{Z}$ si el proceso es discreto (este curso).

* Finalmente, queremos que la  covarianza se conserve bajo todos los "corrimientos de fase" ($\tau$). Es decir

$$ \text{Cov}(X_u, X_v) = \text{Cov}(X_{u + \tau}, Y_{u + \tau}) $$
Si estamos hablando de una serie de tiempo, estamos diciendo que

$$\text{Cov}(X_{t_1}, X_{t_2}) =K_{XX}(t_1, t_2) = K_{XX}(0, t_2-t_1) = K_{XX}(\tau)$$, a esta $\tau$ lo llamamos "corrimiento de fase" (phase shift o simplemete "shift").

Note que pasamos de dos variables $t_1,t_2$ a $0, \tau$ y  a $\tau$ gracias la estacionarieidad. Note tambien que

$$\mathbb{V}(X_t)  = \text{Cov}(X_t, X_t) = K_{XX}(t,t) = K_{XX}(0) = \sigma^2 $$
que es la varianza.

Terminos nuevos: Si la serie no es estacionaria hablamos del termino **heteroscedasticity**, de lo contrario hablamos de **homoscedasticity**. 
Mi traduccion al espanol [**heterosedasticidad**](https://es.wikipedia.org/wiki/Heterocedasticidad)  y **homosedasticidad**.

### Otras notaciones.
Se usa $\gamma_{\tau}$ para definir autocovarianza.  Asumamos $X_t$ con media 0.

$$ \gamma(t_1, t_2) = \mathbb{E}[X_{t_1}, X_{t_2}] = \mathbb{E}[X_t, X_{t+\tau}]$$, donde $\tau=t_2-t_1$ y si es esetacionaria, simplemente,

$$ \gamma(\tau) = \mathbb{E}[X_t, X_{t + \tau}]$$

* En el discreto, asumiendo $X_t=[x_1,  x_2, \cdots, x_n]$ tenemos entonces que

$$\gamma_i = \sum_{j=1}^n X_j X_{i+j} p(x_j) \quad ,  \quad (*)$$

* Si es continuo:

$$ \gamma(\tau) = \int_{-\infty}^{\infty}  X(t) X(t + \tau) p(t) dt$$

Tenemos que poner cuidado a la suma (*), por que nos podemos desbordar.
$i=n$, $i+j >n$, pues $j=1,2, \cdots, n$. Los valores tanto de $i$ como  $i+j$ deben ser menores o iguales a $n$, sino, el computador te genera un error de memoria (podria ser dificil de encontrar).

Para ser mas exactos la formula se debe escribir como

$$ \gamma_i = \sum_{j=1, i+j \le n}^n X_j X_{i+j} p(x_j) = \sum_{j=1}^{n-i} X_j X_{i+j} p(x_j) $$.
En la practica es comun hacer un "padding with zeroes" agregar ceros al final del vector $X$ para poder llegar hasta $n$, los valores entre
$n-i, n$ son todos 0. 

En procesamiento de senhales es comun eliminar el $p()$ y escribir



$$ \gamma_i  = \sum_{j=1}^{n-i} X_j X_{i+j}  $$,
donde se asume que la distribucion es uniforme y se escala por $n$ para eliminar el $1/n$, para el caso discreto.  Para el continuo

$$ \gamma(\tau) = \int_{-\infty}^{\infty}  X(t) X(t + \tau) p(t) dt$$

y a esto lo llaman **correlacion cruzada** o

[cross-correlation](https://en.wikipedia.org/wiki/Cross-correlation)

La idea de la correlacion cruzada es buscar similitudes entre unan senal y la misma mas adelante. O si son dos senales, la correlacion cruzada encuentra simulitudes entre ellas. 




Notacion, se usa notacion $\star$
de la siguiente forma

$(\gamma_{XY})_i = [ X \star Y ]_i = \sum_{j=1}^{n-i} X_j Y_{i+j} $ para el discreto.  Para el continuo

$$ \gamma_{XY}(\tau) = (X \star Y)(\tau) = \int_{-\infty}^{\infty} X(t) Y(t +  \tau) d \tau $$

En el procesamiento de senales se usa $\star$ o $\circledast$ para la correlacion cruzada y $*$ para la convolucion que no la vemos en este curso, pero correponde a las formulas de arriba con $\tau$ cambiado por $-\tau$. 

Terminamos con el concepto de **funcion de correlacion**.

Recordemos

$$ \gamma(t_1, t_2) = \mathbb{E}[ (X_{t_1} - \mu_{X_{t_1}})(Y_{t_2} - \mu_{Y_{t_2}}) ]  $$

La funcion de correlacion se define como

$$ \rho(t_1, t_2) = \frac{\gamma(t_1, t_2)}{\sqrt{{\sigma_{t_1}}^2} \sqrt{\sigma_{t_2}^2}} $$

donde ${\sigma_{t_1}}^2$ y ${\sigma_{t_2}}^2$ son las varianzas de
$X_{t_1}$ y $X_{t_2}$ respectivamente. Recuerde

$$ \sigma_t^2 = \mathbb{E}[ (X_t - \mu_t)^2] $$

Observe que $\rho(t_1, t_2)$ corresponde al coeficiente de correlacion 
de Pearson, de la clase anterior. 

De nuevo, cuando la serie es estacionaria $\sigma_t = \sigma= \sigma_0$
y $\sigma(t_1, t_2)$ depende solo de la diferencia $t_2-t_1 = \tau$, entonces llamamos **funcion de autocorrelacion (**Autocorrelation Function** (ACF). 

$$\rho(\tau) = \frac{\gamma_{\tau}}{\sigma^2} = \frac{\gamma_{\tau}}{\gamma_0} $$

donde $\gamma(\tau)= \gamma(\tau, 0)$. Claramente $\rho(0)=1$. La ACF
es una auto-correlacio  **normalizada**.

Mas formalmente

\begin{eqnarray}
\rho: \mathbb{N} &\to& [-1,1 ] \\
\tau &\mapsto& \rho(\tau) = \rho_{\tau}
\end{eqnarray}
