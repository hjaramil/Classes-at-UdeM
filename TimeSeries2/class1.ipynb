#### Probabilidad. La probabilidad es una funcion del espacio de  muestreo en el intervalo $[0,1]$.  La probabilidad, extricamente hablando, es parte de una teoria llamada **teoria de la medida**. Nosotros en este curso no la enfocamos desde ese punto de vista, que es muy teorico. Esta funcion (probabilidad) mide la posibilidad de que un **evento** se produzca. Un **evento** es una muestra en el espacio de muestreo $S$ (sampling space).

Mas concretamente

\begin{eqnarray}
P : S &\to& [0, 1 ] \\
x &\mapsto& P(x).
\end{eqnarray}

De nuevo, este curso no es  formal, pero probabilidad cumple unas propiedades de la teoria de la medida conocidas como algebra sigma $\sigma$-algebra. 
Ver este [link](https://es.wikipedia.org/wiki/%CE%A3-%C3%A1lgebra), por ejemplo.


Una forma de entender un concepto, es a traves de sus propiedades. 

**Propiedades de la probabilidad**: 

* $0 \le P(x) \le 1$. 0 que nunca ocurre, 1 que ocurre con certeza.
* $P(S)=1$. Recuerde que $S$ es nuestro conjunto de muestreo. Es decir, para nosotros $S$ es el universo.
* Si tienes una secuencia de eventos que son **exclusivos**, $x_i$, $i \in I$
(en un momento explico quien es $I$) entonces

$$ P(\cup_{i \in I} x_i)= \sum_{i \in I} P(x_i) $$
* $P(\phi)=0$. Donde $\phi$ ese el conjunto vacio.

Aca aclaramos los siguientes puntos.
Exclusivos: quiere decir que no pueden ocurrir simultaneamente. Que si se da el uno no se da el otro. Interseccion vacia. Los indicess $I$.
**Notacion**: La letra $I$ la llamo el conjunto de indices. Puede ser finito o no. Por ejemplo $I=\{0,1,2,3\}$, o $I= \mathbb{N}$. , $I = [0,1]$. Este conjunto de indices es infinito no contable y el signo $\sum$ se cambia por el signo $\int_0^1$. Usamos tambien la siguiente notacion.

$$ I_n = \{1,2, \cdots, n \} \quad , \quad n \ge 1. $$
Con esto gano espacio y tiempo en la notacion.
Ejemplo:

$$ \sum_{i=1}^7 x_i = \sum_{i \in I_7} x_i  $$

Ya entendemos el concepto de probabilidad condicional. 
**Independencia**: Dos eventos son independientes si $P(A \cap B)=P(A) P(B)$.

No es muy intiuitiva. Pero podemos reformularla.

$$ P(A | B) = \frac{P(A \cap B)}{P(B)}  \quad (*) $$
$$ P(B | A ) = \frac{P(A \cap B)}{P(A)}  \quad (**) $$


Si $A$ y $B$ son independientes entonces:

$P(A|B) = P(A)$ por que $A$ no depende de $B$ y
$P(B|A) = P(B)$ por que $B$ no depende de $A$
De la ecuacion: (*), y $P(A|B)=P(A)$ tenemos que $P(A \cap B)=P(A) P(B)$. 
igual de la ecuacion (**). Si $P(B|A)=P(B)$ es por que $P(A \cap B)=P(A) P(B)$.
Vemos por que la independencia esta ligada a la probabilidad de la interseccion.

Asumimos que ni $A$ ni $B$ son vacios.

Probemos el teorema de Bayes y lo explicamos.

De las ecuaciones (*) y (**)

$$ P(A \cap B) = P(A | B) P(B) = P(B|A) P(A)$$.

De esta ecuacion

$$P(B|A) = \frac{P(A|B) P(B)}{P(A)} $$

Si el antes (prior) es $A$ y el despues (posterior) es $B$, entonces
podemos entender la probalidad del posterior dado el anterior en funcion
de las probabidades de los eventos y la del anterior dado el posterior
y vice-versa. La probalidad del anterior dado el posterior se puede encontrar
a partir de la probabilidad de los eventos y la del posterior dado el anterior.


Generalicemos: (teoria de conjuntos, leyes de Morgan, etc).

Asumamos que los eventos $B_1, B_2, \cdots, B_n$ son mutuamente exclusivos, no se intersectan y su union es el el universo entonces

$$ A =A \cap ( \cup_{i=1}^n B_i) = \cup_{i=1}^n A \cap B_i$$, como
los eventos $A \cap B_i$ son mutuamente excluyentes. 
De forma que

$$P(A) = \sum_{i=1}^n P(A \cap B_i) = \sum_{i=1}^n P(A | B_i) P(B_i) $$

El teorema de Bayes queda

$$ P(B_i | A) =  \frac{P(A | B_i) P(B_i)}{P(A)} =\frac{P(A | B_i) P(B_i)}{\sum_{i=1}^n P(A | B_i) P(B_i)}$$


Generalicemos la probabilidad de la interseccion de varios conjuntos.
En vez de $A,B$ pensemos $A=A_2$ , $B=A_1$ y escribimos

$$ P(A_2 \cap A_1) = P(A_1) P(A_2 | A_1) $$

Ahora tomemos 3 conjuntos $A_3 \cap A_2 \cap A_1= A_3 \cap (A_2 \cap A_1)$.

$$P(A_3 \cap A_2 \cap A_1) =  P(A_2 \cap A_1) P(A_3 | A_2 \cap A_1) = P(A_1) P(A_2 | A_1)  P(A_3 | A_2 \cap A_1)$$

Por induccion matematica. 

$$ P(A_1 \cap A_2 \cap \cdots \cap A_n) = P(A_1) P(A_2 | A_1) P(A_3 | A_2 \cap A_2) \cdots P(A_n | A_{n-1} \cap A_{n-2} \cdots \cap A_1)  $$

En forma mas compacta (cambiano $\cap$ por $,$, si los eventos fueran independientes

$$ P(A_1, A_2, \cdots, A_n) = \prod_{i=1}^n P(A_i)$$
