## Probabilidad:
No hay espacio para desarrollar la teoria de probabilidades en este curso. Vamos a asumir unos postulados simles y extraer algunas propiedades basicas.
Por definicion la probabilidad es una funcion

\begin{eqnarray}  P: S &\to& [ 0, 1 ]  \\
    x &\mapsto& P(x)  \end{eqnarray}

Los axiomas (postulados)

1. $0 \le P(x) \le 1 $

2. $P(S) = 1$
3. Para una secuencia $x_i$, $i \in I$, donde los eventos $x_i$ son mutuamente exclusivos

$$ P( \cup_{i \in I} x_i)  = \sum_{i \in I} P(x_i)  $$


Con estos axiomas simples se pueden probar una serie de propiedades.

### Probabilidad Condicional:

Matematicamente 

$$P(A | B ) = \frac{P(A \cap B)}{P(B)} $$

Selection_002.png

#### Independencia estadistica

Si $A$ no depende de $B$

$$ P(A | B) = P(A) $$ y
$$ P(B | A) = P(B) $$.

Como mostramos esto. Dos eventos son independientes si $P(A \cap B) = P(A) P(B)$. 

Usando la definicio de probabilidad condicional

\begin{eqnarray}
P(A | B) = \frac{A \cap B}{P(B)} = \frac{P(A) {P(B)}}{P(B)} = P(A) 
\end{eqnarray}


### Teorema de Bayes:
De la defincion de probabilidad condicional tenemos
\begin{eqnarray}
P(A | B) = \frac{A \cap B}{P(B)}   \\
P(B | A) = \frac{A \cap B}{P(A)} 
\end{eqnarray}

De aca
$$ P(A \cap B) = P(B) A(A | B) = P(A) P(B | A)  $$
despejando $P(B|A)$ de la ecuacion de arriba

$$P(B | A ) = \frac{P(A | B) P(B)}{P(A)} $$


## Distribuciones de Probabilidad

Una funcion de distribucion de probabilidad es una funcion (valga la redundancia) donde las salidas (outcomes)  son probabilidades de variables aleatorias. Vamos a ver concretamente. Hay dos tipos de variables (continuas y discretas) y esto se hereda a las distribuciones de probabilidad. Si la variable es discreta la distribcion se llama PMF (Probability Mass Funcion: funcion de masa de probabilidad) y si es continua se llama PDF: Probability Density Distribution (distribucion en densidad de probabilidad)

### Distribucion discreta.
Piense en el universo $S= \cup_{i \in I} x_i $.  Como $S$ es el universo , de la propiedad 4 de la tabla de arriba.

$$P(S) = \sum_{i \in I}  P(X = x_i) = \sum_{i \in I} p_i =1$$
$p_i \ge 0$. $p_i = P(X = x_i)$
$p_i$ es la distribucion discreta de probabilidad. $PMF$.

Si por ejemplo $T \subset S$, con indices $I$ 

$$ P(T) =  \sum_{i \in I} p_i $$

Ejemplo: Un dado el universo es $S=\{1,2,3,4,5,6\}$.
Pensemos en que el dado no tiene sesgo. La probabilidad de cada salia es la misma $1/6$. En este caso la funcion de densidasd de probabilida (PMF) es

$$ p = \frac16$$

Con esto construimos cualquier probabilidad de cualquier subespacio del conjunto de muestreo $S$. Por ejemplo cual es la probabilidad de que tirando un dado salgan los numeros $2  \le x \le 5$ 

$$ P(2 \le X \le 5) = P(X=2 \; o \; X=3, \; o \; X=\; 4  \; o X=5)  = \sum_{i \in I} p_i =\frac{4}{6}= \frac23$$
donde $I={2,3,4,5}$

Vamos a definir algunas distribuciones importantes (continuas o discretas).

1.   La densidad de probabilidad uniforme. Discreto.
Cada evento tiene la misma probabilidad $p=constante$.
Si el espacio tiene $m$ eventos tenemos

$$ P(X = x) = p = \frac1m$$

Si fuera continuo en el intervalo real $[a,b]$

$$ P(X=x) = p =  \frac{1}{b-a}  \int_a^b p(x) dx $$
como $p$ es constante 

$$\frac{1}{b-a} = p  $$


2. La binomial : La variable solo puede tomar dos estados. Ejemplo clasico una moneda. A este tipo de distribciones se les llama Bernoulli. Al experimento se le llama Bernoulli trial (prueba de Bernoulli). Esta prueba tiene varias caracteristicas:

    1. Hay un numero fijo de intentos $n$
    2. Solo tiene dos salidas (caras o sellos)
    3. La probabilidad de evento no depende de los demas.  O sea que si la probalidad de un evento es $p$ la del otro es $q$ y $p+q=1$. 

    Con esta informacion se contruye el modelo matematico de los Bernoulli trials.

$$ P(X=k) = p(k) = \binom{n}{k} p^k p^{n-k} = \binom{n}{k} p^k (1-p)^{n-k} $$

Vamos a hacer una grafica en Python.

Asumamos una moneda sesgada $p=0.6$, $n=10$, las muestras
son $0 \le k \le 10$.







```python
import numpy as np
from scipy.stats import binom
from scipy.stats import norm # no la usamos acas
import matplotlib.pyplot as plt

## set the values
n=10
p=0.6

# el rango de valores
k_values = list( range(n+1))

# pmf: Probability Mass Function
dist = [binom.pmf(k,n,p) for k in k_values]


# graficar
plt.title("binomial con 10 intentos de Bernounili, $p=0.6$, $n=10$")
plt.xlabel("k")
plt.ylabel("P(x=k)")
plt.bar(k_values, dist, color='blue') # esta es la grafica importante
# para salvarlo plt.savefig("binomail2.pdf")
plt.show()




```

 ### Distribucion continua de probabilidades (PDF)

 Aca $x$ varia a lo largo del continuo. Por ejemplo si mi
 dominio es $S=[a,b]$. Tenemos que
 $$ \int_a^b p(x) dx = 1 $$
 podemos en general escribir merjor

 $$\int_{-\infty}^{\infty} p(x) dx $$

 Si queremos la probabilidad en un intervalo $[u,v]$.

 $$ p(u \le x \le v ) = \int_u^v p(x) dx  $$


 Ejemplos:
 1.  La uniforme (ya la hicimos para $S=[a,b]$, pero hamoslo mas general

 $$ \int_S p(x) dx = 1 $$
 Si la probabilidad es uniforme

 $$ \int_S p dx = p \int_S dx = p \mu(S) =1 $$

 $$p = \frac{1}{\mu(S)} $$


 Si $S=[a,b]$, $\mu(S)=b-a$ y tenemos
 $$ p = \frac{1}{b-1} $$

2. La distribucion normal (Gaussiana).

$$ p(x) = \mathcal{N}(\mu, \sigma^2) = \frac{1}{\sigma \sqrt{2 \pi}} \mathrm{e}^{-\frac12 \left ( \frac{x-\mu}{\sigma} \right )^2}  $$

### Razones de importancia para la distribucion normal

1. Muchisimos datos en el continuo tiene la forma de la campana Gaussiana. Por ejemplo estatura de la gente. 

2. Los metodos estadisticos parametricos mas importantes se basan en la districion normal

3. El teorema central del central limite dice que cuando se tienen muchos experimentos con variables aleatorias independientes con media $\mu$ y varianza $\sigma^2$ entonces
en el limite cuando $n \to \infty$ obtenemos la distribucion


$$ Z = \lim_{n \to \infty} \sqrt{n}  \left ( \frac{\overline{X}_n - \mu}{\sigma} \right ) $$

$Z$ es una distribucion normal con media $\mu=0$ y desviacion estandar $\sigma=1$. 


En el caso de la distribucion normal tenemos los siguientes parametros

$$\mu = n p  \quad , \quad \sigma= \sqrt{np (1 - p)} $$

Con esto vamos a ilustrar el teorema central del limite.



```python
import numpy as np
from scipy.stats import binom
from scipy.stats import norm # no la usamos acas
import matplotlib.pyplot as plt

## set the values
n=100
p=0.6

# el rango de valores
k_values = list( range(n+1))

# pmf: Probability Mass Function
dist = [binom.pmf(k,n,p) for k in k_values]

mean = n*p
std = np.sqrt(n*p*(1-p))

distribution = norm(mean, std)
probabilities = [ distribution.pdf(v) for v in k_values]


# graficar
plt.title("binomial con 10 intentos de Bernounili, $p=0.6$, $n=10$")
plt.xlabel("k")
plt.ylabel("P(x=k)")
plt.bar(k_values, dist, color='blue') # esta es la grafica importante
plt.plot(k_values , probabilities, color='red', label="Normal distribtion")
# para salvarlo plt.savefig("binomail2.pdf")
legend = plt.legend(loc=' upper right', fontsize='x-small')
plt.show()
```

Otro punto interesante de la distribcuion normal es el punto de minimos cuadrados


El problema de minimos cuadrados resulta, por ejemplo, 
cuando queremos ajustar $A \theta = b$, donde $A$ es matriz y $x,b$ vectores. Podemos pensar en los vectores $x_i \in \mathbb{R}^n$, $i=1,2, \cdots , m$. Queremos ajustar los puntos de la mattiz $A$ ( los $x_i$ son filas de la matriz $A$)

Gauss dijo: Que cada linea de esa ecuacion

$$ \epsilon_i = x_i^T \theta - b_i$$
Dijo mas, que esos errores son independientes y tiene una distribucion PDF $\phi(\epsilon_i)$. 

Resulto con analisis que

$$\phi(\epsilon_i) = \frac{1}{\sqrt{2 \pi \sigma^2}} \mathrm{e}^{-\epsilon_i^2/(2 \sigma^2)} $$

Cuando los eventos son independientes la probabilidad combianada es el producto de las probabilidades individuales.
(Likelihood: verosimilitud)


\begin{eqnarray*}
      P = \prod_{i=1}^m \phi(\epsilon_i) =  \frac{1}{\sqrt{ (2 \pi)^m \sigma^{2m}}}
      \mathrm{e}^{- \frac{1}{\sigma^2} \sum_{i=1}^m (x_i \theta - b_i)^2} 
      =   \frac{1}{\sqrt{ (2 \pi)^m \sigma^{2m}}}
      \mathrm{e}^{- \frac{1}{\sigma^2} \| A \theta - b \|^2} .
\end{eqnarray*}

Maximizar el likelihood es equivalente a minizar el exponente. Es decir:


$$ \min_{\theta} \| A \theta - b \|^2 $$ 


### Cumulative Distribution Function:  Las funciones de distribucion acumuladas.

Por definicion las acumuladas se escriben

$$ F_X (x) = P( X \le x )$$ .$. 

Si es discreta escribimos:

$$ F_X(x) = \sum_{x_k \le x} p(x_k) $$

Si la variable aleatoria $X$ es continua

$$ F_X(x) = \int_{\infty} ^{x}  p(t ) dt$$

Podemos el teorema fundamental del calculo.

$$ p(x) = \frac{d F_X(x)}{dx}  $$


En este sentido

$$\lim_{x \to -\infty} {F_X}(x) = 0 \quad , \quad \lim_{x \to \infty} F_X(x)  = 1 \, .$$

Las funciones de distribucion acumuladas son muy importantes pero dificiles de evaluar.

Por ejemplo para la Gaussina

$$ F_x(x) = \int_{-\infty}^x \frac{1}{\sigma \sqrt{2 \pi}} \mathrm{e}^{-\frac12 \left ( \frac{t-\mu}{\sigma} \right )^2} $$
