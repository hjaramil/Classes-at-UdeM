## Estadisticas acerca de las series A(p)
### Media

El modelo AR(p) es

$$ X_t = c + \sum_{i=1}^p \phi_i X_{t-i} + Z_t
 $$

 $$ \mu = \mathbb{E}[ c ] + \sum_{i=1}^p \mathbb{E} [\phi_i X_{t-i} ] + \mathbb{E}[{Z_t}]$$

De aca

$$\mu = c = \sum_{i=1}^p \phi_i \mu + 0 $$

Despejamos $\mu$

\begin{eqnarray}
\mu - \mu \sum_{i=1}^p \phi_i = c \\
\mu= \frac{c}{1-\sum_{i=1}^p \phi_i}
\end{eqnarray}



## cual es la varianza $\sigma^2$ en un modelo AR(p)?

No se la respuesta. Pero vamos hacer cositas.

Ausmoasmos $c=0$,

$$ X_t = \sum_{i=1}^p \phi_i X_{t-i}  + Z_t $$

Tomamos la esperanza de $X_t X_{t-k}$ 

\begin{eqnarray}
\mathbb{E}[ X_t X_{t-k}] &=& \sum_{i=1}^p \mathbb{E}[X_{t-i} X_{t-k}]  + \mathbb{E} [ Z_t X_{t-k}]  \\
\gamma_k &=& \sum_{i=1}^p \phi_i \gamma_{k-i} + \mathbb{E}[Z_t X_{t-k} ] 
\end{eqnarray}

Donde

\begin{eqnarray}
\mathbb{E}[Z_t X_{t-k} ] = \left \{
    \begin{array}{cc}
    \sigma_Z^2 & \text{si  } k = 0 \\
    0  & \text{si  } k =1,2, \cdots 
    \end{array}
    \right .
\end{eqnarray}

Encontramos las siguientes ecuaciones:

$$ \sigma^2 = \gamma_0 = \sum_{i=1}^p \phi_i \sigma_{k-i} + \sigma_Z^2 \quad (1) $$
$$ \gamma_k = \sum_{i=1}^p \phi_k \gamma_{k-i}$$


Realmente no tenemos solucionado el problema de la varianza en una serie AR(p). Por ejemplo (de la ecaucion (1) 

$$\sigma^2 = \gamma_0 = \phi_1 \gamma_1 + \phi_2 \gamma_2 + \cdots + \phi_{\mathrm{p}} \gamma_{\mathrm{p}} + \sigma_Z^2 \quad (2) $$

Uno diria que como $\rho_k= \gamma_k/\gamma_0$ , $\gamma_k= \rho_k \gamma_0$. Reemplazando en (2)

$$\sigma^2 = \gamma_0 = \phi_1 \rho_1 \gamma_0 + \phi_2 \rho_2 \gamma_0 + \cdots + \phi_{\mathrm{p}} \rho_{\mathrm{p}} \gamma_0 + \sigma_Z^2 \quad (2) $$

Despejamos $\gamma_0 = \sigma^2$.

\begin{eqnarray}
\gamma_0 = \sigma^2 = \frac{\sigma_Z^2}{1 - \phi_1 \rho_1 - \phi_2 \rho_2 - \cdots - \phi_{\mathrm{p}} \rho_{\mathrm{p}}}
\end{eqnarray}



## Yule-Walker Equations para un modelo AR(p)

La idea de esto es encontrar los $\phi_i$ , $i=1,2, \cdots, p$ a
partir de los datos . Este problema es un problema de *inversion*. En estadistica se usa la palabra *estimacion*

El modelo:  ($c=0$)

$$ X_t = \sum_{i=1}^p \phi_i X_{t-i} + Z_t $$

Hay varias formas de estimar los $\phi_i$ ( Wienner-Hopf usa
minimos cuadrados, lo discutimos mas adelante. Preferido de los
que usan digital signal analysis) pero Yule-Walker es el preferido por los estadisticos.

Se toman correlaciones con distintos corrimientos de la senal y se crea un sistema lineal de $m$ ecuaciones con $m$ incognitas.

* Lag 1: Calculamos

\begin{eqnarray} X_t X_{t+1} &=& X_t \left ( \sum_{i=1}^{\mathrm{p}} \phi_i X_{t-i+1} + Z_t  \right ) \\
&=& \sum_{i=1}^{\mathrm{p}} \phi_t X_t  X_{t-i+1} + X_t Z_{t+1}
\end{eqnarray}
Tomamos la esperanza a ambos lados.


$$\mathbb{E}[X_t X_{t+1}] =\gamma_1 = \sum_{i=1}^{\mathrm{p}} \phi_i \mathbb{E}[ X_t X_{t-i+1}] + \mathbb{E}[ X_t Z_{t+1}] $$

O sea 

$$ \gamma_1 = \sum_{i=1}^{\mathrm{p}} \phi_i \gamma_{i-1} $$
Dividmos todo por $\gamma_0$

$$ \rho_1 = \sum_{i=1}^{\mathrm{p}} \phi_i \rho_{i-1}   \quad (3)$$



* Lag 2
\begin{eqnarray} X_t X_{t+2} &=& X_t \left ( \sum_{i=1}^{\mathrm{p}} \phi_i X_{t-i+2} + Z_t  \right ) \\
&=& \sum_{i=1}^{\mathrm{p}} \phi_t X_t  X_{t-i+2} + X_t Z_{t+2}
\end{eqnarray}
Tomamos la esperanza a ambos lados.


$$\mathbb{E}[X_t X_{t+2}] =\gamma_1 = \sum_{i=1}^{\mathrm{p}} \phi_i \mathbb{E}[ X_t X_{t-i+2}] + \mathbb{E}[ X_t Z_{t+2}] $$

O sea 

$$ \gamma_2 = \sum_{i=1}^{\mathrm{p}} \phi_i \gamma_{i-2} $$
Dividmos todo por $\gamma_0$

$$ \rho_2 = \sum_{i=1}^{\mathrm{p}} \phi_i \rho_{i-2}   \quad (4) $$

* Lag $k$
\begin{eqnarray} X_t X_{t+k} &=& X_t \left ( \sum_{i=1}^{\mathrm{p}} \phi_i X_{t-i+k} + Z_t  \right ) \\
&=& \sum_{i=1}^{\mathrm{p}} \phi_t X_t  X_{t-i+k} + X_t Z_{t+k}
\end{eqnarray}
Tomamos la esperanza a ambos lados.


$$\mathbb{E}[X_t X_{t+k}] =\gamma_k = \sum_{i=1}^{\mathrm{p}} \phi_i \mathbb{E}[ X_t X_{t-i+k}] + \mathbb{E}[ X_t Z_{t+k}] $$

O sea 

$$ \gamma_k = \sum_{i=1}^{\mathrm{p}} \phi_i \gamma_{i-k} $$
Dividmos todo por $\gamma_0$

$$ \rho_k = \sum_{i=1}^{\mathrm{p}} \phi_i \rho_{i-k}  \quad (5) $$

Asi hasta el 

* Lag p

\begin{eqnarray} X_t X_{t+\mathrm{p}} &=& X_t \left ( \sum_{i=1}^{\mathrm{p}} \phi_i X_{t-i+k} + Z_t  \right ) \\
&=& \sum_{i=1}^{\mathrm{p}} \phi_t X_t  X_{t-i+\mathrm{p}} + X_t Z_{t+k}
\end{eqnarray}
Tomamos la esperanza a ambos lados.


$$\mathbb{E}[X_t X_{t+\mathrm{p}}] =\gamma_k = \sum_{i=1}^{\mathrm{p}} \phi_i  \mathbb{E}[ X_t X_{t-i+\mathrm{p}}] + \mathbb{E}[ X_t Z_{t+\mathrm{p}}] $$

O sea 

$$ \gamma_{\mathrm{p}} = \sum_{i=1}^{\mathrm{p}} \phi_i \gamma_{i-k} $$
Dividmos todo por $\gamma_0$

$$ \rho_{\mathrm{p}}= \sum_{i=1}^{\mathrm{p}} \phi_i \rho_{i-\mathrm{p}}   \quad (6) $$

De las ecuaciones: (3) , (4), (5), (6) observamos el sistuiente sistema de ecuaciones con incognitas:


\begin{eqnarray}
\begin{pmatrix}
\rho_0 & \rho_1 & \rho_2 & \cdots & \rho_{\mathrm{p}-1} \\
\rho_1 & \rho_0 & \rho_1 & \cdots & \rho_{\mathrm{p}-2} \\
\vdots &  &  &   \cdots & \vdots \\
\vdots &  &  &   \cdots & \vdots \\
\rho_{\mathrm{p}-2} & \rho_{\mathrm{p}-3} & \cdots & \cdots & \rho_1 \\
\rho_{\mathrm{p}-1} & \rho_{\mathrm{p}-2}  & \cdots & \cdots & \rho_0 
\end{pmatrix}
\begin{pmatrix}
\hat{\phi}_1 \\
\hat{\phi}_2 \\
\vdots \\
\vdots \\
\hat{\phi}_{\mathrm{p}-1} \\
\hat{\phi}_{\mathrm{p}} 
\end{pmatrix} = 
\begin{pmatrix}
\rho_1 \\
\rho_2 \\
\vdots \\
\vdots \\
\rho_{\mathrm{p}-1} \\
\rho_{\mathrm{p} -2 } 
\end{pmatrix}
\end{eqnarray}

Como $\rho_0=1$.


\begin{eqnarray}
\begin{pmatrix}
1 & \rho_1 & \rho_2 & \cdots & \rho_{\mathrm{p}-1} \\
\rho_1 &1 & \rho_1 & \cdots & \rho_{\mathrm{p}-2} \\
\vdots &  &  &   \cdots & \vdots \\
\vdots &  &  &   \cdots & \vdots \\
\rho_{\mathrm{p}-2} & \rho_{\mathrm{p}-3} & \cdots & \cdots & \rho_1 \\
\rho_{\mathrm{p}-1} & \rho_{\mathrm{p}-2}  & \cdots & \cdots & 1 
\end{pmatrix}
\begin{pmatrix}
\hat{\phi}_1 \\
\hat{\phi}_2 \\
\vdots \\
\vdots \\
\hat{\phi}_{\mathrm{p}-1} \\
\hat{\phi}_{\mathrm{p}} 
\end{pmatrix} = 
\begin{pmatrix}
\rho_1 \\
\rho_2 \\
\vdots \\
\vdots \\
\rho_{\mathrm{p}-1} \\
\rho_{\mathrm{p} -2 } 
\end{pmatrix}
\end{eqnarray}

Este es un sistema de p ecuaciones con p incognitas. La matriz
\begin{eqnarray}
R = \begin{pmatrix}
1 & \rho_1 & \rho_2 & \cdots & \rho_{\mathrm{p}-1} \\
\rho_1 &1 & \rho_1 & \cdots & \rho_{\mathrm{p}-2} \\
\vdots &  &  &   \cdots & \vdots \\
\vdots &  &  &   \cdots & \vdots \\
\rho_{\mathrm{p}-2} & \rho_{\mathrm{p}-3} & \cdots & \cdots & \rho_1 \\
\rho_{\mathrm{p}-1} & \rho_{\mathrm{p}}  & \cdots & \cdots & 1 
\end{pmatrix}
\end{eqnarray}

Es simetrica, positiva definida y Toeplitz. 

De forma que

$$\Phi = R^{-1} \rho $$

donde

\begin{eqnarray}
\rho = \begin{pmatrix} \rho_1 \\ \rho_2 \\ \vdots \\ \rho_p 
\end{pmatrix} 
\end{eqnarray}

Otra forma de estimar los $\phi_i$ es con minimos cuadrados.

$X_t = \sum_{i=1}^{\mathrm{p}} \phi_i X_{t-i} + Z_t$

Tendo datos, los datos son $Y_t$ y mis datos obedecen al modelo
de arriba. Queremos hallar los $\phi_i$ tales que 

$J(\phi) = \| Y_t - X_t \|^2 = \sum_{i=1}^m (y_i - x_i)^2$  sea minimo
donde $X_t= (x_i)$, $Y_t=(x_i)$ 
Demos hallar $\phi$ tal que $J(\phi)$ sea minimo.

Toca $\nabla J(\phi) = 0$. Donde el gardiente sea 0, se obtiene $\phi$. Tarea (el metodo de Wiener-Hofp).

Ejemplo numerico


```python
# Yule-Walker Equations:
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# parametros
phi1=0.8
phi2=-0.3
phi3=0.1
# el sistema es estable (converge) por que las raices del
# polinomio caracteristico estan por fuera del circulo unitario

# generamos el ruido
np.random.seed(1)
mu, sigma = 0, 1
N=1000
Z= np.random.normal(mu, sigma, N)

X=np.zeros(N)
X[0] = Z[0]
X[1] = Z[1]
X[2] = Z[2] 

for t in range(3, N):
    X[t] = Z[t] + phi1*X[t-1] + phi2*X[t-2] + phi3*X[t-3]
```

```python
plt.xlabel("numero de muestra")
plt.ylabel("valor de la muestra")
plt.title("proceso AR(3)")
plt.plot(X)
plt.show()
```

```python
# creamos el sistema Yule-Walker.
M=3
Rho=np.zeros((M,M))

# calculamos los rho (ACF). Podemos hacerlo con muchos metodos
# pero usamos statsmodels.
rho = sm.tsa.stattools.acf(X)[0:M+1]

rhovec = rho[1:M+1]

# construccion de la matriz

for i in range(M):
    for j in range(i, M):  # triangular superior
        Rho[i,j] = rho[j-i]  # los 1 incluidos
        # el signo !=  signica no-igual, o diferente, otros lenguajes usan <> 
        if i != j : Rho[j,i]=Rho[i,j]  # triangular inferior

# print(Rho)

# usamos el algebra lineal de NumPy.
phivec = np.linalg.solve(Rho, rhovec)

print("phi0=%.1f, phi1=%.1f, phi2=%.1f"%(phivec[0], phivec[1], phivec[2]))
phivec
```

```python
# usamos la libreria de Yule-Walker de statsmodels
rhoYK , sigma = sm.regression.yule_walker(X, order=3, method="mle")
rhoYK
```

# Model Selection Part I: PACF (Partial AutoCorrelation Function)

Repaso de conceptos imortantes que vamos a necesitar.

Recordamos el coeficiente de correlacion Pearson entre dos variables aleatorias $X$ y $Y$.

$\rho_{XY} = \frac{Cov(X-\mu_X,Y-\mu_y)}{\sqrt{\mathbb{V}(X-\mu_x)} \mathbb{V}(Y-\mu_Y)} $

Asumamos que los datos tienen una distribcion uniforme que por el momento estan centrados, o sea que las medias son 0.

Matematicamente el problema es el siguiente:

Tenemos una serie de puntos en $\mathbb{R}^2$
$(x_1, y_1), (x_2, y_2), \cdots , (x_m, y_m)$ 
Por alguna razon sabemos que la relacion entre estos puntos es lineal , o sea que $y_i = \theta_0 + \theta_1 x_i$ Y que si no fuera asi es por que hay ruido.


Quisieramos encontrar la linea que se ajusta a los puntos con el minimo posible error (least squares).

La linea tiene dos parametros $(\theta_0, \theta_1)$ (el intercepto y la pendiente.

Si quisieramos ajustar todos los puntos a esta linea recta tendriamos el siguiente sistema de ecuaciones:


\begin{eqnarray}
y_1 &=& \theta_0 + x_1 \theta_1 \\
y_2 &=& \theta_0 + x_2  \theta_1 \\
&\vdots& \\
y_m &=& \theta_0 + x_m \theta_1 \\
\end{eqnarray}

Llevemos esto a una forma matricial:

$$ y = x \theta $$ donde

\begin{eqnarray}
y = \begin{pmatrix}
y_1 \\ y_2 \\ \vdots \\ y_m
\end{pmatrix} \quad , \quad
x= \begin{pmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots & \vdots \\
1 & x_m 
\end{pmatrix}
\quad , \quad
\theta = \begin{pmatrix} \theta_0 \\ \theta_1 \end{pmatrix}
\end{eqnarray}

El sistema $y=x \theta$ no tiene solucion, en general, por que
tiene $2$ columnas y muchisimas filas $m>2$. Que hacemos?

Una opcion es *minimos cuadrados*. Minimizar la suma de los cuadrados de la diferencia. Podemos definir una funcion de costo (objetivo)

$$ Error(\theta_0, \theta_1, x, y ) = \frac12  \| y - x \theta \|^2
= \frac12  \sum_{i=1}^m (y_i - \theta_0 - \theta_1 x_i)^2$$

Queremos minimizar la funcion de costo .
Se toma el gradiente $\nabla_{\theta} Error = 0$. Se halla $\theta$ para que esto suceda.

Se puede ver , suando cualculo multivariado 

$$ \nabla_{\theta} Error  = x^T x \theta - x^T y$$

Como queremos que el gradiente sea 0 obtnemos las *ecuaciones normales*

$$x^T x \theta = x^T y  $$
Este sistema es mas facil de resolver. Por ejemplo
la matrix $x^T x$ es $2 \times 2$ y es, en general (si no fuera, deberiamos regularizar) invertible (no singular).

En este caso

$$ \theta =  (x^T x)^{-1} x^T y $$

Una forma simple de obtener las ecuaciones normales.

\begin{eqnarray}
x \theta &=& y \\
x^T x \theta &=& x^T y  
\end{eqnarray}
