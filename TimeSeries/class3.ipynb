## Cross-covariance y autocovariance (covarianza cruzada y autocovarianza)

Ya definimos en la clase anterior covarianza $C(X,Y)$.
Definamos la covarianza cruzada.

$$K_{XY}(t_1, t_2) = Cov(X_{t_1}, Y_{t_2}) =E[(X_{t_1} - \mu_{X_{t_1}}) (Y_{t_2} - \mu_{Y_{t_2}})] = E[X_{t_1} Y_{t_2}] - \mu_{X_{t_1}} \mu_{Y_{t_2}} $$

Vamos a cambiar $t_1, t_2$ por otra notacion mas conveniente.  $t_1=t$, $t_2 = t_1 + \tau$, $\tau=t_2 - t_1$

$$K_{XY}(t, t+\tau) = Cov(X_{t}, Y_{t+\tau}) =E[(X_{t} - \mu_{X_{t}}) (Y_{t+\tau} - \mu_{Y_{t+ \tau}})] = E[X_{t} Y_{t}] - \mu_{X_{t}} \mu_{Y_{t+ \tau}} $$

Si $X=Y$ estamos hablando de Autocovarianza
$$ K_{XX}(t, \tau) = E[(X_t - \mu_X)(X_{t+\tau} - \mu_Y] = E[X_t X_{t+\tau}] - \mu_{X_t} \mu_{Y_{t+\tau}} $$



Hablemos mas tel termino "weak stationary"
Un proceso es weak stationary si.

* El primero momento (la esperanza) es contante. Es decir $E[X_t] = E[X_{t + \tau}]=\mu$ para todo $\tau \in \mathbb{Z}$ si es discreta. Si es continua $\tau  \in \mathbb{R}$.
* El segundo momento (varianza) tambien es independiente todo "shift" . Mas concretamente $V[X_t] = V[X_{t+\tau}]$
* La covarianza cruzada tambien es invariante bajo shifts.

$$ Cov(X_t, X_{t_2})  =K_{XX}(t_1, t_2) = K_{XX}(0, t_2-t_1) = K_{XX}(\tau) $$

$$Var(X_t) = Cov(X_t, X_t) = K_{XX}(t,t) = K_{XX}(0) = \sigma^2 $$

Este es el tipo de estacionaridad que vamos a asumir en este curso.

Definimos nuevos terminos. (shift=desplazamiento).

Funcion de autocorrelacion no normalizada

$$ \gamma(t_1, t_2) = E[X_{t_1} X_{t_2}] $$
asumamos que $X$ tiene media 0.

Mas concretamente. Si $X_t$ es discreta , estacionaria y el "shift" es $\tau$

$$ \gamma_{\tau} = \sum_{j=1}^n X_j X_{j + \tau} p(x_j) $$

Si $X_t$ continua

$$ \gamma(\tau) = \int_{-\infty}^{\infty} X(t) X(t + \tau) p(t) dt  $$

$$ \gamma_{\tau} = \sum_{j=1}^n X_j X_{j + \tau} p(x_j) $$

Si en tu computador $X_j$ es un arreglo de $n$ posiciones,
entonces piense en $j=n$ y $\tau > 0$. Para $\tau>0$ y $j=n$ esta expresion arroja error de memoria (memory out of bounds) en el computador.

Como se evita este error.

$$ \gamma_{\tau} = \sum_{j=1, \tau+j \le n}^n = \sum_{j=1}^{n-\tau} X_j X_{j+ \tau} p(x_{\tau}) $$

Hay un campo que se llama DSP: Digital Signal Processing que tiene relacion con lo que estamos viendo.

En DSP se usa la expresion correlacion cruzada (cross-correlation)

$$ [\gamma_{XY}]_i = \frac1n \sum_{i=1}^{n-i} X_j Y_{i+j}  $$
Si $X=Y$, es autocorrelacion
$$ [\gamma_i] = \frac1n \sum_{i=1}^{n-i} X_j X_{i+j}  $$

En convolucion:

$$ convolucion = \frac1n \sum_{i=1}^{n-i} X_j X_{i-j}  $$

Notaciones:

Correlacion 
$$ (\gamma_{XY})_i = [X \star Y] = \sum_{j=1}^{n-i} X_j Y_{i+j}  $$


En el continuo

$$\gamma_{XY}(\tau) = \int_{-\infty}^{\infty} X(t) Y(t + \tau) dt  $$

La cross-correlation  function no normalizada puede variar entre $-\infty$ , $\infty$ queremos una normalizacion.

$$ \gamma(t_1, t_2) =  E[(X_{t_1} - \mu_{X_{t_1}})(Y_{t_2} - \mu_{Y_{t_2}})]$$

normalizamos

$$ \rho(t_1, t_2) = \frac{\gamma(t_1, t_2)}{\sqrt{\sigma_{t_1}^2} \sqrt{\sigma_{t_2}^2}} = \frac{\gamma(t_1, t_2)}{| \sigma_{t_1} | | \sigma_{t_2} | }$$


Realmente a nosotros **en este curso** solo nos interesan los **auto**. 

Entonces definimos (estacionarias)

$$ \rho(\tau) = \frac{\gamma_{\tau}}{\gamma_0} = \frac{\gamma_{\tau}}{\sigma^2}  $$

$\sigma^2 = \gamma(0)$. 
A esto es lo que en el curso llamaremos ACF: Auto-correlation function. Funcion de autocorrelacion.

Claramente $\rho(0)=1$.

## Procesos Autoregressivos.
Autoregressive Process (AR)

### Motivacion.
Voy a traer cosas de DSP (Digital Signal Processing)

Los filtros lineales son convolutivos. Es decir si se tiene
una respuesta impulso (discreta, de ahora en adelante) $h_i$,
la entrada (input, datos) son $x_i$ y la salida es $y_i$.
Entonces el filtro que resulta de "filtrar" los datos $x_i$ con la respuesta impulos $h_i$ es:  (lineal)

$$ y_i = \sum_{k=1}^p x_{i} h_{i-k} = (x*h)_i $$

Aparece de nuevo la **convolucion** en el contexto .

Un proceso intersante es hallar el $h_i$ de datos $y_i$
a este proceso se le llama en DSP: **deconvolucion**

Vamos a ver algo asi pero en el gremio de time series se hace un proceso llamado Yule-Walker.

Existen unos filtros que hacen un "delay" retraso en la senal.

Por ejemplo:

$$ y_i = x_{i+j} = \sum_{k=1}^p x_{i-k} h_k  \quad , \quad j  \ge 1 $$

este un proceso que en DSP se llama predictivo. Hallar los $h_k$ se llamaria predictive deconvolution.

Un caso mas particular. Es hallar el hoy a partir del ayer y mas para atras.

$$  X_{i} = c + \sum_{k=1}^p  \phi_k X_{i-k}  + Z_i   \quad  (1) $$
$c$ es constante (que se llama **drift**)


A este proceso se le llama **autoregression** de orden $p$
y se nota AR(p).

Un ejemplo particular. 
**Random Walk**

$$X_t = X_{t-1} + Z_t$$


Si hacemos una comparacion de este modelo con el modelo vemos el random walk es un modelo autoregresivo con $p=1$ y $\phi_1 = 1$. $c=0$.

Comenzamos por que mi metodo es ir de lo simple a lo complejo con $p=1$. Lo llamaremos AR(1)

$$ X_t = c + \phi_1 X_{t-1} + Z_t = c + \phi X_{t-1} + Z_t $$
El rudio $Z_t$ tiene media 0 y variance $\sigma^2_Z$

Calculemos estadisticas de $X_t$ basados en la suposicion de estacionaridad debil.

**OJOOOOOOO* : el simbolo $X_t$ puede representar la muestra $t$ de la secuencial $(X_1, X_2, \cdots, X_t, \cdots) $, de acuerdo al contexto $X_t$ puede
representar toda la secuencia en conjunto
$X_t = (X_1, X_2, \cdots, X_t, \cdots) $ que
es $X_t$ sin desplazamiento  o shift $=0$.
$X_{t+\tau}$ tambien puede representar la muestra
 $t+\tau$ de la secuencia $X_t$ (con desplazamiento $\tau$. 

*  Media: 

\begin{eqnarray}
\mu = E(X_t) &=& E[ c + \phi X_{t-1} + Z_t ] \\
&=& E[c] + \phi E[X_{t-1}] + E[Z_t] \\
&=& c  + \phi \mu + 0 
\end{eqnarray}

De forma que
$$\mu =  \frac{c}{1-\phi}$$.

Comentarios:
* Si $c=0$ la media $\mu=0$.
* que pasa si $\phi=1$? El proceso no converge. O sea que el Random Walk **no es estacionario**
Vamos a ver (falta la varianza) que $|\phi| < 1$.
Por ahora con $\phi \ne 1$ es suficiente para que 
sea estacinaria en la media.

Mas sobre el drift $c$.

Si el proceso es estacionario $c=\mu(1-\phi)$.

Normalizacion en la media es restar la media para que la nueva media sea 0.

$$X_t \leftarrow  X_t - \mu $$
Para evitar conflictos y ambiguedasdes

$$ \hat{X}_t = X_t - \mu $$ 

$$X_t = c + \phi X_{t-1} + Z_t = \mu(1 -\phi) + \phi X_{t-1} + Z_t $$
Restemos $\mu$ a ambos lados

$$X_t - \mu =  \phi( X_t - \mu) + Z_t $$

$$\hat{X}_t = \phi \hat{X}_t + Z_t $$  este es AR(1) con el mismo $\phi$. Libre del drift $c$. 

Por esta razon, en general, vamos a asumir que $c=0$.

* Variance: 

\begin{eqnarray}
\sigma^2 &=& Var(X_t) = Var(c) + Var(\phi X_{t-1}) + Var(Z_t)  \\
&=& 0 + \phi^2 Var(X_{t-1}) + \sigma_Z^2  \\
&=& \phi^2 \sigma^2  + \sigma_Z^2 
\end{eqnarray}

Despejamos $\sigma^2$.

\begin{eqnarray}
\sigma^2 = \frac{\sigma_Z^2}{1 - \phi^2} 
\end{eqnarray}

Comentarios:
* Como $\sigma^2$ es positivo entonces
$1 - \phi^2 > 0$, $| \phi | < 1$. Condicion necesaria para que la serie sea estacionaria.

* $c$ no cuenta. 

* Hay un error en esta formula. Se corrije en la proxima clase.
