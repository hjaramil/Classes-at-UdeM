 Mostrar que el backshift en una constante es la misma constnate.

 $$B c = B (1)^t c = (1)^{t-1}c = c$$

La solucion del sistema homogeneo es la combinacion de estas dos

$$X_{th} =C_0 \left (\frac{1}{r_1} \right )^t +  C_1 \left ( \frac{1}{r_2} \right )^t $$

Se necesitan dos condiciones iniciales. $X_0=X_0,  X_1=X_1$


Para $t=0,1$

\begin{eqnarray}
X_0 &=& C_0 + C_1 \\
X_1 &=& \frac{C_0}{r_1} + \frac{C_1}{r_1}
\end{eqnarray}

Lo dejo como ejercio (esta en las notas)

\begin{eqnarray}
C_1 = \frac{r_1 ( X_0 - r_2 X_1) }{r_1 - r_2 } \quad , \quad 
C_2 = \frac{r_2 (r_1 X_1 - X_0)}{r_1 - r_2 }
\end{eqnarray}




Que tal si las raices son repetidas?

Comencemos con un caso sencillo

$$ (1-B)^2 X_t = 0$$

En este caso $p(B)=(1- B)^2$ donde $B=1$ es una raiz doble.
Las souciones (como en ecuaciones diferenciales ordinarias)

$X_t = C_j t^j$ ,  $t=0,1$.
Probemos esto (dos soluciones linealmente independientes del homogeneo)
Las soluciones son $C_0,  C_1 t$.

$$(1 - B)^2 (C_0) = (1-B)(1-B) C_0 = (1-B) (C_0 - C_0) = (1-B)0=0 $$

$$(1-B)^2 (C_1 t) = (1-B) (1-B) C_1 t =(1-B) (C_1 t - C_1(t-1)) =(1-B)(1) = 1 - 1 = 0 $$

Para las condiciones iniciales (X_0, X_1) conocidas, tenemos que

\begin{eqnarray}
C_0 = X_0 \\
C_1 = X_1 
\end{eqnarray}

Ahora consideremos un caso mas general


$$ (1 - \varphi B)^2 X_t = 0 $$

Proponemos dos soluciones $X_t = t^j \varphi^j$, $j=0,1$.
Probemoslo

*  Si $j=0$ 

    $X_t = \varphi^t$ 

* Si $j=1$ (es mi tarea)
\begin{eqnarray}
(1 - \varphi  B)^2 (t \varphi^t) &=& (1 - \varphi B)(1 - \varphi B)(t \varphi^{t})  \\
&=& (1 - \varphi B) (t \varphi^t - \varphi (t-1) \varphi^{t-1}) \\
&=& (1 - \varphi B) (t \varphi^t - t \varphi^t + \varphi^t) \\
&=& (1 - \varphi B)(\varphi^t) \\
&=& \varphi^t - \varphi \varphi^{t-1}  = 0.
\end{eqnarray}


Tenemos dos soluciones linealmente independientes


* $X_{th0} = C_0 \varphi^t $
* $X_{th1} = C_1 t \varphi^t$

La general homogenea es:

$$ X_{th} = C_0 \varphi^t + C_1 t \varphi^t  $$


La general del problema es la particular mas la homogenea


$$ X_t = X_{th} + X_{tp}  $$


Consideremos ahora el problema general de dos raices diferences




Hallemos $C_1$ , $C_2$

\begin{eqnarray}
X_0 &=& C_1 + C_2 + Z_0  \quad (1) \\
X_1 = \frac{C_1}{r_1} + \frac{C_2}{r_2}  + Z_1 + \frac{1}{r_1 - r_2}\left ( \frac{r_1}{r_2} - \frac{r_2}{r_1} 
\right ) Z_0 &=&  \frac{C_1}{r_1} + \frac{C_2}{r_2} + Z_1 +  \frac{r_1 + r_2 }{r_1 r_2 } Z_0
\end{eqnarray}

Eliminacion Gaussiana

\begin{eqnarray}
 X_0 - r_1 X_1 &=& C_2 + Z_0 -  \frac{r_1 C_2}{r_1} - r_1 Z_1 - \frac{r_1 + r_2}{r_2} Z_0  \\
 C_2 &=& \frac{X_0 - r_1 X_1 - Z_0 + r_1 Z_1 + \frac{r_1 + r_2}{r_2}}{1 - r_1/r_2}  \\
 &=& \frac{r_2 X_0 - r_1 r_2 X_1 - r_2 Z_0 + r_1 r_2 Z_1 + r_1 Z_0 + r_2 Z_0}{r_2 - r_1 } \\
 &=& \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 Z_1 + r_1 Z_0}{r_2 - r_1 }
\end{eqnarray}

Hallemos $C_1$ de la (1)

\begin{eqnarray}
C_1 &=& X_0 - Z_0 - C_2  \\
&=& X_0 - Z_0 -  \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 Z_1 + r_1 Z_0}{r_2 - r_1 } \\
&=& \frac{r_2 X_0 - r_1 X_0 - r_2 Z_0 + r_1 Z_0 - r_2 X_0 + r_1 r_2 X_1 - r_1 r_2 Z_1 - r_1 Z_0 }{r_2 - r_1} \\
&=& \frac{-r_1 X_ - r_2 Z_0 + r_1 r_2 X_1 - r_1 r_2 Z_1}{r_2 - r_1 }
\end{eqnarray}




Enconrramos entonces que:

\begin{eqnarray}
C_1 &=& \frac{-r_1 X_ - r_2 Z_0 + r_1 r_2 X_1 - r_1 r_2 Z_1}{r_2 - r_1 } \\
C_2 &=& \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 Z_1 + r_1 Z_0}{r_2 - r_1 }
\end{eqnarray}

De forma que la solucion es:



\begin{eqnarray*}
      X_t &=&   C_1 \left ( \frac{1}{r_1} \right )^t + C_2   
      \left ( \frac{1}{r_2} \right )^t \\  && 
      \frac{1}{r_1 - r_2} 
      \left [ (r_1 - r_2) Z_t +  \left (r_1 \frac{1}{r_2} - 
        r_2 \frac{1}{r_1} \right ) Z_{t-1} \right .   \\
        && \left . +  \left (r_1 \frac{1}{r_2^2} - r_2 \frac{1}{r_1^2} \right ) Z_{t-2}  
        \cdots + 
        \left (r_1 \frac{1}{r_2^n} - r_2 \frac{1}{r_1^n} \right ) Z_{t-n} + \cdots 
      \right ]  \\
  C_1 &=& \frac{-r_1 X_0 - r_2 Z_0 + r_1 r_2 X_1 - r_1 r_2 Z_1}{r_2 - r_1 } \\
C_2 &=& \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 Z_1 + r_1 Z_0}{r_2 - r_1 }    
\end{eqnarray*}

Como hay causalidad (en el computador no hay indices negativos)

\begin{eqnarray}
X_0 &=& C_1 + C_2 + Z_0  \\
X_1 &=& \frac{C_1}{r_1} + \frac{C_2}{r_2} + Z_1 +  \frac{r_1 + r_2 }{r_1 r_2 } Z_0 \\
&\vdots & \\
X_n &=& C_1 \left ( \frac{1}{r_1} \right )^n + C_2   
      \left ( \frac{1}{r_2} \right )^n \\  && 
      \frac{1}{r_1 - r_2} 
      \left [ (r_1 - r_2) Z_n +  \left (r_1 \frac{1}{r_2} - 
        r_2 \frac{1}{r_1} \right ) Z_{n-1} \right .   \\
        && \left . +  \left (r_1 \frac{1}{r_2^2} - r_2 \frac{1}{r_1^2} \right ) Z_{t-2}  
        \cdots + 
        \left (r_1 \left ( \frac{1}{r_2} \right )^n - r_2  \left (\frac{1}{r_1} \right )^n \right ) Z_{t-n} + \cdots 
      \right ]  
\end{eqnarray}

Esto es valido para culquier par de raices complejas $r_1, r_2$.
Sin embargo vamos a simplificar esta expresion usando numeros complejos.

Notas en los complejos: $z$ 
* $z=a + \mathrm{i} b $, $a, b \in \mathbb{R}$ 
* Forma polar $z= \rho \mathrm{e}^{i \theta} = \cos \theta + \mathrm{i} \sin \theta  $ donde $\rho=\sqrt{a^2 +  b^2 }$, $\theta= \arctan(b/a)$.

La ecuacion estrella es la caracteristica $p(B)=1 - \phi_1 B - \phi_2 B^2$.
Como las soluciones son complejas el discriminante es negativo, es decir $\phi_1^2 - 4 \phi_2 < 0$.  Las soluciones se pueden escribir como


$$ \varphi_{\pm}  = \frac{\phi_1  \pm \mathrm{i} \sqrt{- \phi_1^2 - 4 \phi_2}}{2}$$

Identifico $\varphi_1 = \varphi_{+}$, $\varphi_2 = \varphi_{-}$.
Entonces

$$ (1 - \phi_1 B - \phi_2 B^2)=(1 - \varphi_1 B)(1 - \varphi_2 B)$$
y de aca sabemos que

$$ r_1 = \frac{1}{\varphi_1} \quad , \quad r_2 = \frac{1}{\varphi_2} $$

Racionalizamos las raices:

$$r_1 = \frac{2}{\phi_1 + \mathrm{i} \sqrt{ - \phi_1^2 - 4 \phi_2}} = \frac{2 \left ( \phi_1 - \mathrm{i} \sqrt{-\phi_1^2 - 4 \phi_2} \right )}{\phi_1^2 + ( -\phi_1^2 -  \phi_2)} = -\frac{2  \left ( \phi_1 - \mathrm{i} \sqrt{-\phi_1^2 - 4 \phi_2} \right )}{4 \phi_2} $$

Podemos escribir las partes real e imaginarias separadas.

$$r_1 = -\frac{\phi_1}{2 \phi_2} + \mathrm{i} \frac{\sqrt{-\phi_1^2 - 4 \phi_2}}{2 \phi_2 } = a + \mathrm{i} b  $$
donde

$$ a =  -\frac{\phi_1}{2 \phi_2}  \quad  , \quad b= \frac{\sqrt{-\phi_1^2 - 4 \phi_2}}{2 \phi_2 } $$

En forma porlar tenemos que:

$$ r_{\pm} = \rho \mathrm{\rho}^{\pm \mathrm{i} \theta} $$


donde
$$ \rho=\sqrt{a^2 + b^2} \quad ,\quad \theta = \arctan \left ( \frac{b}{a} \right ) $$

Las dos raices en forma polar son:
$$ r_1 = \rho \mathrm{e}^{\mathrm{i} \theta}   \quad , \quad r_1 = \rho \mathrm{e}^{-\mathrm{i} \theta} $$


Simplificamos asumiendo $X_0=Z_0$, $X_1=Z_1$. Entonces:



\begin{eqnarray}
  C_1 &=& \frac{-r_1 X_0 - r_2 X_0 + r_1 r_2 X_1 - r_1 r_2 X_1}{r_2 - r_1 } = \frac{r_1 + r_2 }{r_1 - r_2 } X_0 \\
C_2 &=& \frac{r_2 X_0 - r_1 r_2 X_1 + r_1 r_2 X_1 + r_1 X_0}{r_2 - r_1 }    = -\frac{r_1 + r_2}{r_1 - r_2} X_0
\end{eqnarray}

Variable compleja: Con $r_1=\rho \mathrm{e}^{\mathrm{i} \theta} $ y
$r_2 = \rho \mathrm{e}^{-\mathrm{i} \theta}$ entonces


$$ r_1 + r_2 = 2 \rho \cos \theta  
\quad , \quad r_2 - r_2 = 2 \mathrm{i} \rho \sin \theta $$

De forma que

\begin{eqnarray}
C_1 = \frac{\cos \theta}{\mathrm{i} \sin \theta} = -\mathrm{i} \cot \theta \quad , \quad   C_2  = \mathrm{i} \cot \theta 
\end{eqnarray}


\begin{eqnarray}
C_1 \left ( \frac{1}{r_1} \right )^n + C_2   
      \left ( \frac{1}{r_2} \right )^n &=& -\mathrm{i} \cot \theta \rho^{-n} \mathrm{e}^{-\mathrm{i} n \theta } + \mathrm{i} \cot \theta \rho^{-n} \mathrm{e}^{\mathrm{i} \theta} \\
      &=& \mathrm{i} \rho^n \cot \theta ( \mathrm{e}^{\mathrm{i} n \theta} - \mathrm{e}^{- \mathrm{i} n \theta} ) \\
      &=& \mathrm{i} \rho^n \cot \theta  (2 \mathrm{i} \sin n \theta) \\
      &=&-2 \rho^{n} \cot \theta \sin n \theta) 
\end{eqnarray}

Ahora bien 

\begin{eqnarray}

      &=& \frac{1}{2 \mathrm{i} \rho \sin \theta} \left [ 2 \mathrm{i} \rho \sin \theta Z_{n} + (\mathrm{e}^{2 \mathrm{i} \theta}  - \mathrm{e}^{-2 \mathrm{i} \theta} ) Z_{n-1}  +

## Estadisticas:

Recuerde del proceso AR(2):

$$ X_t = c + \phi_1 X_{t-1} + \phi_2 X_{t-2} + Z_t $$

Comencemos:

* *media* : 

$$ \mu = E[X_t] = E[c] + \phi_1 E[X_{t-1}] + \phi_2 E[X_{t-2}] + E[Z_t]  = c + \phi_1 \mu  + \phi_2 \mu$$

Despejamos $\mu$.

$$ \mu = \frac{c}{1 - \phi_1 - \phi_2 } $$
Claramente si el dritf $c=0$, la media es 0 y $1 \ne \phi_1 + \phi_2$.

* *varianza*:     
De $X_t = \phi_1 X_{t-1} + \phi_2 X_{t-2} + Z_t $

\begin{eqnarray}
\gamma_k = E[X_t X_{t-k}] &=& \phi_1 E[X_{t-1} X_{t-k} ] + \phi_2 E[X_{t-2} X_{t-k}] + E[Z_t X_{t-k]} ]\\
&=& \phi_1 \gamma_{k-1} +  \phi_2 \gamma_{k-2} + E[Z_t X_{t-k}]
\end{eqnarray}


Donde:

\begin{eqnarray}
E[Z_t X_{t-k}]= \left \{ \begin{array}{2}
\sigma_Z^2 & \text{si } k = 0\\
0 &  k = 1,2, \cdots, 
\end{array} \right .
\end{eqnarray}

De forma que
\begin{eqnarray}
\sigma^2 &=& \gamma_0 = \phi_1 \gamma_1 + \phi_2 \gamma_2 + \sigma_z^2 \quad , \quad k=0  \quad (3)  \\
\gamma_k &=&  \phi_1 \gamma_{k-1} +  \phi_2 \gamma_{k-2} \quad (2)
\end{eqnarray}

$\sigma^2$ depende de $\gamma_1$ y $\gamma_2$ pero no sabemos todavia que valor tienen estos.

\begin{eqnarray}
\gamma_1 = \phi_1 \gamma_0 + \phi_2 \gamma_{-1} &=&\phi_1 \sigma^2 + \phi_2 \gamma_1  \\
\gamma_1 &=& \frac{\phi_1 \sigma^2 }{1 - \phi_2}
\end{eqnarray}

De la ecuacion (2) para $k=2$. 
$$\gamma_2 = \phi_1 \gamma_1 + \phi_2 \gamma_0 = \phi_1 \frac{\phi_1 \sigma^2 }{1 - \phi_2} + \phi_2 \sigma^2 = \sigma^2 
\left (  \frac{\phi_1^2  + \phi_2(1-\phi_2) }{1 - \phi_2}\right )   $$


De la ecuacion (3):

\begin{eqnarray}
\sigma^2 &=& \phi_1 \gamma_1 + \phi_2 \gamma_2 + \sigma_z^2 
= \phi_1 \frac{\phi_1 \sigma^2 }{1 - \phi_2} + \phi_2 \sigma^2 
\left (  \frac{\phi_1^2  + \phi_2(1-\phi_2) }{1 - \phi_2}\right ) + \sigma_Z^2
\end{eqnarray}

Pasando a restar a la izquierda y factorizando $\sigma^2$ tenemos:




\begin{eqnarray}
\sigma^2 \left [ 1 -\frac{\phi_1^2  }{1 - \phi_2} -  \phi_2 
\left (  \frac{\phi_1^2  + \phi_2(1-\phi_2) }{1 - \phi_2}\right )  \right ]  = \sigma_Z^2 
\end{eqnarray}

De aca se despejar $\sigma^2$

Simplifiquemos el coeficiente de $\sigma^2$.

\begin{eqnarray}
1 -\frac{\phi_1^2  }{1 - \phi_2} -  \phi_2 
\left (  \frac{\phi_1^2  + \phi_2(1-\phi_2) }{1 - \phi_2}\right ) = \frac{1 - \phi_2 - \phi_1^2 -\phi_2 \phi_1^2 - \phi_2^2 + \phi_2^3}{1- \phi_2} \\
= \frac{(1 - \phi_2^) - \phi_1^2( 1 + \phi_2) + \phi_2(\phi_2^2 -1)}{1 - \phi_2} \\
= \frac{(1-\phi_2)(1 + \phi_2) - \phi_1^2(1 + \phi_2) - \phi_2(1 - \phi_2)(1 + \phi_2) }{1 - \phi_2} \\
= \frac{(1 + \phi_2) ( 1 - \phi_2 - \phi_1^2 - \phi_2(1 - \phi_2}{1 - \phi_2} \\
= \frac{ (1 + \phi_2)( 1 - 2 \phi_2 + \phi_2^2 - -\phi_1^2 }{1 - \phi_2} \\
= \frac{(1 + \phi_2)( (1 - \phi_2)^2 -  \phi_1^2 )}{1 - \phi_2} \\
= \frac{(1 + \phi_2)(1 - \phi_2 - \phi_1 )(1 - \phi_2 + \phi_1}{1 - \phi_2}
\end{eqnarray}

De forma que

\begin{eqnarray}
\sigma^2 = \frac{(1 - \phi_2 )\sigma_Z^2}{(1 + \phi_2)(1 - \phi_2 - \phi_1 )(1 - \phi_2 + \phi_1)}
\end{eqnarray}
