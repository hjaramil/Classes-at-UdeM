## Procesos autoregresivos de orden 1 AR(1)

$$X_t = c + \phi_1 X_{t-1} + Z_t$$
$c$ se llama "drift".

Asumamos que el proceso es estacionario.

Vamos a hallar las estadisticas (atributos) de esta serie de tiempo.


* Mean:
    \begin{eqnarray}
    \mu = E(X_t) &=& E[c + \phi X_t + Z_t] \\
    &=& E[c] + \phi E[X_t] + E[Z_t] \\
    &=& c + \phi \mu + 0 \\
    &=& c + \phi \mu \\
    \mu - \phi \mu &=& c \\
    (1 - \phi) \mu = c
    \end{eqnarray}

Despejamos $\mu$.

$$ \mu = \frac{c}{1 - \phi}  $$

Observaciones:
* Si $c=0$ la media es 0.
* $\phi \ne 1$. De hecho vamos a probar que $| \phi |<1$. Esto cuando trabajemos con la varianza.

### normalizacion por la media. 
Justificacion de por que algunas veces esta bien asumir que $c=0$.

A la serie $X_t$ le vamos a restar la media

$$ X_t \leftarrow X_t - \mu$$

Para evitar confusion
$\tilde{X_t} = X_t - \mu $

Reemplazamos $c $ por $c=\mu(1- \phi)$ en la la ecuacion AR(1)

$$X_t = c + \phi_1 X_{t-1} + Z_t = \mu(1 - \phi) + \phi X_{t-1}+ Z_t $$

Esto es, restando $\mu$.

$$ X_t - \mu = \phi(X_{t-1} - \mu) + Z_t$$
En terminos de la varable con virgulilla.

$$\tilde{X_t} = \phi \tilde{X}_{t-1} + Z_t$$


* Varianza: Asumamos $c=0$, entonces

\begin{eqnarray} \sigma^2 = Var[X_t] &=& Var[\phi X_{t-1}] + Var[Z_t]+ 2 Cov[\phi X_{t-1} Z_t]  \\
&=& \phi^2 Var[X_{t-1}] + \sigma_Z^2 + 2 {\phi Cov[X_{t-1}, Z_t] } \\
&=& \phi^2 \sigma^2 + \sigma^2_Z 
\end{eqnarray}
Despejando $\sigma^2$

$$\sigma^2 = \frac{\sigma_Z^2}{1- \phi^2} $$

Observaciones:
* $\sigma^2> 0 , \sigma^2_Z> 0 > 0$, $1 - \phi^2 > 0$

$1 - \phi^2 > 0$, $\phi^2 < 1 $, $ | \phi | < 1$.



#### Funcion de autocorrelacion de AR(1)
Asumamos que $c=0$, $X_t = \phi X_{t-1} + Z_t$

$$\gamma_1 = Cov(X_t, X_{t+1})= E[ X_t (\phi X_t + Z_t)] = E[ \phi X_t^2 + X_t Z_{t+1}] = \phi E[X_t^2 + E[X_t Z_{t+1}] = \pi Var[X_t]  $$

Hagamos lo mismo para otro "lag" $\tau$ (lag=shift=desplazamiento)

\begin{eqnarray}
X_{t + \tau} &=& \phi X_{t+\tau -1 } + Z_{t + \tau} \\
X_t X_{t+ \tau} &=& \phi X_t X_{t+ \tau-1} + X_t Z_{t + \tau} \\
E[X_t X_{t+ \tau} ] &=& \phi E[X_t X_{t+\tau - 1} ]+ E[X_t Z_{t + \tau}] \\
\gamma_1 &=& \phi \gamma_{\tau-1} 
\end{eqnarray}

Entonces:

\begin{eqnarray}
\gamma_0 &=& \sigma^2 \\
\gamma_1 &=& \phi \gamma_0 = \phi \sigma^2 \\
\gamma_2 &=& \phi \sigma_1 = \phi^2 \sigma^2 \\
&\vdots &  \\
\gamma_{\tau} &=& \sigma^2 \phi^{\tau} \quad , \quad \tau=0,1, \cdots, 
\end{eqnarray}
Como $\gamma_{\tau}$ es simetrico...

Que pasa si $\tau < 0$?
Si $tau$ es menor que 0

En vez de $X_{t + \tau}$ seria $X_{t - \tau}$.
y siguiendo el mismo proceso

$\gamma_{\tau} = \phi{-\tau} sigma^2 $,

la que siempre funciona es
$\gamma_{\tau} = \phi^{| \tau |} \sigma^2$
$\rho_0 = 1$.
Si $\phi<0$ la funcion oscila, 

la funcion de autocorrelacion ACF: AutoCorrelation Funcion es la normalizacion de la funcion de autocovarianza $\gamma_j$

La ACF es:

$$ \rho_j = \gamma_j/\gamma_0 = \phi^{| \tau |}  $$

Diferentes formas de hallar una solucion analitica cerrada de AR(1).

Recuerde $X_t = c + \phi X_{t-1} + Z_t $.
Simplifiquemos asumiento que $c=0$.

#### Backsubstitution: Sustitucion hacia atras:

\begin{eqnarray}
X_t &=& \phi X_{t-1} + Z_t \\
&=& \phi ( \phi X_{t-2} + Z_{t-1}) + Z_t = \phi^2 X_{t-2} + \phi Z_{t-1} + Z_t \\
&=& \phi^2 (\phi X_{t-3} + Z_{t-2}) + \phi Z_{t-1} + Z_t = \phi^3 X_{t-3} + \phi^2 Z_{t-2} + \phi Z_{t-1} + Z_t  \\
&\vdots& \\
&=& \phi^k X_{t-k} + \sum_{j=0}^{k-1} \phi^j Z_{t-j}
\end{eqnarray}


De aca:
$$ \phi^k X_{t-k} = X_t - \sum_{j=0}^{k-1} \phi^j Z_{t-j}  $$
Sera que podemos tomar esta suma hasta el infinito?

Vamos tomar la varianza en ambos lados y tomar al limite al infinito.


\begin{eqnarray}
\lim_{k \to \infty} E \left [ \left ( X_t - \sum_{j=0}^{k-1} \phi^j Z_{t-j} \right)^2
    \right ] = \lim_{k \to \infty} \phi^{2 k} E[X_{t-k}^2]
\end{eqnarray}

Para que el limite exista $| \phi | < 1$. 
En ese caso el limite es 0, 

tenemos que
$$ X_t = \sum_{j=0}^{\infty} \phi^j Z_{t-j} $$










En la practica no podemos tomar series al infinito.
En la practica

$X_t = \sum_{j=0}^N \phi^t Z_{t-j} $

Quiero causalidad. Es decir (los computadores son causales) hay una **primera** muestra $Z[0]$.
En la programacion (Python, Fortran,....)
Los indices en Python , C, C++ , Java comienan en 0.
$t-j \ge 0$.

$$ X_t = \sum_{j=0}^N \phi^t Z_{t-j} \quad , \quad t=1,2, \cdots , N \quad , j \le t $$

Tambien se puede hacer **Forward Substitution**
(tarea).

\begin{eqnarray}
X_1 &=& \phi X_0 + Z_0 \\
X_2 &=& \phi X_1 + Z_1 = \phi (\phi X_0 + Z_0) +Z_1 = \phi^2 X_0 + \phi Z_0 + Z_1 \\
&\vdots&
\end{eqnarray}


#### Backshift operator (operador the desplazamiento en reversa).

El backshift operator toma un termino y lo mueve un lugar hacia atras.

$$ B(X_n) = X_{n-1} $$

$$B^2 (X_n) = B (B (X_n)) = B(X_{n-1}) = X_{n-2}$$.
Y por induccion

$$ B^k(X_n) = X_{n-k} $$

Escribimos el sistema AR(1) usando el operador B

\begin{eqnarray}
X_t &=&  \phi X_{t-1} + Z_t \\
X_t - \phi X_{t-1} &=& Z_t \\
X_t - \phi B X_t &=& Z_t \\
(1 - \phi B)X_t &=& Z_t
\end{eqnarray}

Postulamos que

$$X_t = \sum_{j=0}^{\infty} \varphi_j Z_{t-j} = \varphi(B) Z_t =  \quad (1) $$
donde $\varphi(B)= \sum_{j=0}^{\infty} \varphi_j B^j$. 

$$X_t = \varphi(B) Z_t = \varphi(B)(1 - \phi B)X_t \quad (2) $$

Vamos a hallar los coeficientes $\varphi_j$ comparando termino a termino aquellos $X_t$ de la ecuacion (2). 

\begin{eqnarray}
1 &=& (1 + \varphi_1 B + \varphi_2 B^2 + \cdots + \varphi_j B^j + \cdots ) ( 1 - \phi B) \\
&=& 1 + \varphi_1 B + \varphi_2 B^2 + \cdots + \varphi_j B^j + \cdots  - \phi B - \varphi_1 \phi B^2 - \varphi_2 \phi B^3 - \cdots - \varphi_{j-1} \phi B^j + \cdots  \\
&=& 1 + (\varphi_1 - \phi) B + (\varphi_2 - \varphi_1 \phi)B^2 + \cdots + (\varphi_j -  \varphi_{j-1} \phi)  B^j + \cdots 
\end{eqnarray}

Piense que $1 = 1 + 0 B + 0 B^2 + \cdots 0 B^j + \cdots $


Termino a termino:

\begin{eqnarray}
\varphi_0 &=& \phi^0 = 1 \\
\varphi_1 &=&\phi  \\
\varphi_2  &=& \varphi_1 \phi = \phi^2 \\
&\vdots &
\varphi_j = \varphi_{j-1} \phi = \phi^j 
\end{eqnarray}

Encontramos entonces (nuevamente)

$$X_t = \sum_{j=0}^{\infty} \phi^j Z_{t-j} = \sum_{j=0}^{\infty} \phi^j B^j Z_t $$

Quiero recurrir a los productos y cocientes notables del algebra.




$$(1 - \phi B) (1 + \phi B + \phi^2 B^2 + \cdots + \phi^{n-1} B^{n-1}) =  (1 - \phi^n B^n) $$

De otra forma

$$  \frac{1 - \phi^n B^n}{1 - \phi B}= \sum_{i=0}^{n-1} \phi^i B^i $$

Si $| \phi | < 1$ podemos tomar el limite cuando $n \to \infty $ y obtener la **serie geometrica**

$$ \frac{1}{1 - \phi B}  = 1 + \phi B + \phi^2 B^2 + \cdots + \phi^j B^j + \cdots = \sum_{j=0}^{\infty} \phi^j B^j$$

Para que me sirve eso.
El problema que queremos resolver es

$$ (1 - \phi B) X_t = Z_t$$
de forma ingenua despejamos $X_t$

$$ X_t = \frac{Z_t}{1 - \phi B}= \frac{1}{1 - \phi B} Z_t
=  \sum_{j=0}^{\infty} \phi^j B^j Z_t =\sum_{j=0}^{\infty} \phi^j Z_{t-j}  $$

### Ecuaciones en diferencias
Que son ecuaciones en diferencias (difference equations).
Una ecuacion en diferencias de orden $n$ es

$$a_0 x_t + a_1 x_{t-1} + \cdots + a_n x_{t-n} = f_n  $$

Los coeficientes $a_i$ son numeros reales, $f_n$ es un termino fuente (forcing term, source)
y la incognita es $x_t$. Se necesitan (como las ecuaciones diferenciales ordinarias) $n$ condiciones iniciales.

En nuestro caso particular tenemos

$$X_t = c + \phi X_{t-1} + Z_t $$
o sea

$$X_t - \phi X_{t-1} = c + Z_t  $$
el termino fuente (source o forcing term) es $f=c+Z_t$. 

Comenzamos con la facil, la homogenea $f=0$.

$$ X_t - \phi X_{t-1} = 0  $$

Como la resolvemos. Pensemos en la recursion

\begin{eqnarray}
X_1 &=& \phi X_0 \\
X_2 &=& \phi X_1 = \phi^2 X_0 \\
&\vdots &  \\
X_t &=& \phi^t X_0
\end{eqnarray}

Volvamos al backshift operator $B$.

$$ (1 - \phi B) X_t = 0 $$.
Vamos a defir el polinomio
$$ p(B) = 1 - \phi B$$ como el polinomio caracteristico 
del sistema.
La raiz de este polinomio caracteristico $P(r)=0$.
$r=1/\phi$. La solucion del proceso es

$$ X_{th} = C \left ( \frac1r \right )^t $$.
Toca determinar la constante (en un momento).
Veriquemos que esta solucion es la correcta para el AR(1).

$$ (1 - \phi B) X_{th} = X_t - \phi X_{t-1} =  C \left ( \frac1r \right )^t - \phi  C \left ( \frac1r \right )^{t-1} = C \left (  \left ( \frac1r \right)^t -\left ( \frac1r \right)^t  \right ) =0 $$


Quen es $C$? Se usan las condiciones iniciales. Asumamos que conocemos a $X_0$. Ahora, para $t=0$

$$ X_0 = C \left ( \frac1r \right)^0 =C  $$

O sea que $C=X_0$ y la solucion es

$$ X_{th} = \left ( \frac1r \right)^t X_0 = \phi^t X_0 $$

El problema que estamos resolviento

$$ (1 - \phi B) X_t = c + Z_t $$

Lo interesante es que cualquier solucion a la ecuacion homogenea aporta a la solucion de este problema. 
Prueba: En general asumar $L x = f$ donde $L$ es un operador lineal. Si $x_{h}$ es solucion a la homogenea $L x_h = 0$.

Ahora la solucion general es $x= x_h + x_p$ donde
$L x_p = f$. 

$L( x_h + x_p) = L x_h + L x_p = 0 + L x_p = f$.

Hallemos la solucion particular.

$$ (1 - \phi B) X_t = c + Z_t$$.

Recuerden la serie geometrica:

$$\frac{1}{1- \phi B} = 1 + \phi B + \phi^2 B^2 + \cdots + \phi^n B^n  + \cdots  $$

La solucion particula es

$$  X_{tp} = (1 + \phi B + \phi^2 B^2 + \cdots + \phi^n B^n + \cdots + ) ( c + Z_t) $$



La solucion general es:

\begin{eqnarray}
 X_t  &=&  X_{th} + X_{tp} 
 &=& \phi^t X_0  + c \sum_{i=0}^{\infty} \phi^i + \sum_{i=0}^{\infty} \phi^i Z_{t-i} 
\end{eqnarray}

(tarea $B c = c$ para toda constante $c$. 


#### Yule-Walker Equations.
Hasta el momento solo hecho modelacion (modeling) . Es decir dados los $\phi$ y el $Z_t$ y condiciones iniciales hallar $X_t$ .El problema inverso es dado $X_t$ , $Z_t$ y las condiciones iniciales hallar $\phi$

$$ X_t = \phi X_{t-1} + Z_t$$.


\begin{eqnarray}
X_t X_{t+1} &=& X_t ( \hat{\phi} X_t + Z_{t-1}) \\
&=& \hat{\phi} X_t X_t + X_t Z_{t+1} 
\end{eqnarray}

Calculo la autocovarianza $\gamma_1$ .

\begin{eqnarray}
\gamma_1 &=& E[ X_t X_{t+1}  ] = E[ \hat{\phi} X_t X_t + X_t Z_{t+1} ] \\
&=& \hat{\phi} E[X_t X_t] + E[X_t Z_{t+1}] \\
&=& \hat{\phi} E[X_t X_t]
\end{eqnarray}
y dividiendo por $\gamma_0=\sigma^2$.

$$\rho_1 = \hat{\phi} \rho_0  $$

De aca que

$$ \hat{\phi} = \frac{\rho_1}{\rho_0} $$

Ejemplo: Asuma el proceso AR(1)

$$ X_t = 0.3 X_{t-1} + Z_t $$
donde $Z_t$ es ruido blanco con media 0, varianza 1.













```python
import numpy as np
import matplotlib.pyplot as plt




# ruido blanco Z
np.random.seed(1)  # por que ?
mu, sigma = 0, 1
N=1000
Z = np.random.normal(mu, sigma, N)

# parametros AR(1)
phi=0.3
X=np.zeros(N)
X[0]=Z[0]

for t in range(1,N):
    X[t] = Z[t] + phi*X[t-1]



# grafica
plt.xlabel("sample number")
plt.ylabel("sample value")
plt.title("AR(1), $\phi=0.3$")
plt.plot(X)
# plt.savefig("mi figura en pdf")
plt.show()


```

$X_t =  \phi^t X_0  + \sum_{i=0}^{\infty} \phi^i Z_{t-i} $

```python
## implementacion de la forma cerrada analitica. Closed form analytic.
X2 = np.zeros(N)
X2[0] = Z[0]

for t in range(1,N):
    X2[t] = phi**t*X2[0] + sum( phi**i  *Z[t-i] for i in range(t))

    # grafica
plt.xlabel("sample number")
plt.ylabel("sample value")
plt.title("AR(1), $\phi=0.3$")
plt.plot(X2)
# plt.savefig("mi figura en pdf")
plt.show()


```
