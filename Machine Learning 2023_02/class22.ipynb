# Artificial Neural Networks (ANN).

Varios puntos.

* Las ANN resuelven el problema universal de ajustar una funcion "arbitaria" usando capas con nodos conectados (redes neuronales: definicion mas adelante). Cuando digo arbitria es
$f: \mathbb{R}^n \to \mathbb{R}^m.$ que satisface casi todo lo que las funciones en la practica satisfacen. Que sea integrable.
Solo funciones muy raras no se ajustan a esto y no tiene sentido practico. Las ANN son
[universal approximators](https://en.wikipedia.org/wiki/Universal_approximation_theorem).

* La solucion de un problema usando regresion lineal polinomica para resolver imagenes es inviable. Esto lo vemos usando lo que ya estudiamos. Empezamos hoy, terminamos en la proxima clase.

Piense en una imagen, por ejemplo de $100 \times 100$ pixeles.
La imagen puede ser a color con tres canales RGB(Red-Green-Blue), o $HSV$ (Hue-saturated-value), o CMYK (Cian-Mayenta-Yellow-Black).
Las primeras tienen 3 canales la ultima 4. Por simplicar pongamosla en blanco y negro. 1 solo canal.
Pregunta. Si usamos polinomios para regresion lineal (polinomial) cuantos terminos necesitamos? cuantos $\theta_i$. Cada pixel es una variable (feature). Hay 10000 pixeles. Piense en que hay $n$ pixeles, el polinomio de grado $2$ (el mas pequenho no lineal esta dado por


$$ \theta_0 + \theta_1 x_1 + \cdots + \theta_n x_n + \theta_{n+1} x_1 x_2 + \cdots + \theta_m x_{n-1} x_n + \theta_{m+1} \theta_1^2 + \theta_{m+2} x_2^2 + \cdots + \theta_q x_n^2 . \tag{1} $$



Contar los $\theta$'s es contar $q$.  El conteo es $q+1$.

Asuma las variables $1, x_1, x_2, \cdots , x_n$ y polinomios con coefecientes (de los $\theta)$ de la forma

$$ 1^{i_0} x_1^{i_1} \cdots x_n^{i_n} .$$
Como estamos hablando de polinomios de grado dos, los exponentes suman dos de la siguiente forma.


$i_0 + i_1 + \cdots + i_n =2$.
Veamos que este modelo simula todos terminos del polinomio (1)

Por ejemplo:
* $\theta_0$ se obtiene haciendo $i_1=i_2 = \cdots =i_n=0$, $i_0=2$.
* $\theta_n x_n$ se obtiene haciendo $i_n=1$, $i_0=1$.

* El termino $\theta_{m+1} x_1^2$ se obtiene con  $i_0=1$, $i_1=2$
y asi sucesivamente.

Como contamos todos esos terminos. Recuerde la ecuacion para esto coteo


$$ \binom{n+k-1}{k} .$$
Esta cuenta $x_0 + x_1 + x_2 + \cdots x_n = k$

$$ \binom{n+k}{k} = \frac{(n+1)!}{k! n!}= \frac{(n+1)(n+2) \cdots (n+k)}{k!} .$$

Verifiquemos que esta ecuacion funciona.
Polinomio en dos variables cuadratico.

$$p(x_1, x_2) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \theta_3 \theta_1 \theta_2 + \theta_4 x_1^2 + \theta_5 x_2^2 .$$


Cuantos terminos tiene el polinomio? 6.

Verifiquemos la formula para el conteo de este modelo
$n=2$, $k=2$

$$\binom{n+k}{k} =\binom{4}{4} = \frac{4!}{2! 2!}= \frac{4 \times 3 \times 2!}{2! 2!}= 6 .$$

En nuestro problema hablamos de 10000 (100 x 100) pixeles

Si contamos los terminos de un polinmio de grado 2 (el mas pequenho no lineal) encontramos
$n=10000$, $k=2$.


$$\binom{10002}{1000} = \frac{10001 \times 10002}{2}
= 50015001.0 .$$




Si el polinomio fuera cubico tenemos

$$\binom{10003}{3} = \frac{10003!}{10000! 3!}= \frac{(10003)(10002)(10001}{6} = 166.766.685.001.0
 .$$
 Este numero es del orden de $10^{12}$. Ahora bien recuerde el problema de regresion $X \theta = y$,
 la matriz $X$ tiene $m \times n$ entradas, pero
 $\theta$ tiene $n$ entradas y en general $m>n$,
 o sea que la matriz $X$ tiene mas de $10^{22}$ entradas. Y esto es totalmente inadmisible para los computadores de hoy en dia.

## Historia. Solo dos bullets.
* 1943:
[threshold logic](https://es.wikipedia.org/wiki/Neurona_de_McCulloch-Pitts)
Warren McCulloch (neurofisiologo) Walter Pitts (matematico). Se juntaron y crearon el primer modelo, muy simple, de redes neuronales. Se basaron en circuitos electricos logicos (y, o, nor, xor) para simular el comportamiento del cerebro dado un "umbral".

* 1958:
[perceptron](https://en.wikipedia.org/wiki/Perceptron).
Frank Rosenblatt, en el Cornell Aeronautical Laboratory, introdujo este concepto que lo explicamos enseguida y que se generaliza al MLP: Multilayer Perceptron que es la arquitectura de las redes neuronales modernas.

Un ejemplo, bobito, de la neurona de McCulloc and Pitts es el siguiente.

Asuma que usted estudia  ir a un paseo

Hay una lista de datos $(x)$ para donde va
* Distancia al lugar de destino
* mode de desplazamiento: carro, a pie, bicicleta, avion.
* clima del lugar de destino.
* humedad del lugar de destino.
* costo de alojamiento y avion/barco/etc.

De otro lado estas los pesos (weights, $\theta$)
(entre 0 y 1, 0 para nada y 1 para mucho)

* Que tanto me gusta desplazarme largas distancias
* Que tanto me gusta caminar. Me dan miedo los aviones?
* Que tanto me gusta los climas frios, humedos, calientes
* Cuanto dinero tengo, o hasta cuanto quiero gastar.

A lo ultimo su decision depende de un promedio ponderado $\sum \theta_i x_i$ y si este promedio rebasa cierto numero, voy o sino no.

En el "nor" los pesos ($\theta$) son $w_1 = w_2 = -1$

El ```nor``` quiere decir que solo es verdadero cuando ninguno de los dos es verdadero (ni el uno, ni el otro).



El operador ```xor```. Este operador marca un hito en las redes neuronales dado que es imposible simularlo con una red neuronal de una sola capa, como la McCuloch and Pitts.

El ```xor``` es un "o" exclusivo. Es decir **solo si uno  de los dos es verdad** el resultado es verdadero.
