# The Pseudoinverse (La pseudoinversa)
El problema mas comun en en sistemas lineales es $Ax = b$,
donde $A \in \mathbb{R}^{m \times n}$, $x \in \mathbb{R}^n$ y $b
\in \mathbb{R}^m$. Si $A$ tiene inversa, claramente $x = A^{-1}b$.
Pero general $A$ no tiene inversa.

Si las columnas de $A$ son linealmente independientes  $A^T A$ tiene inversa y $x = (A^T A)^{-1} A^T b $, a
$(A^T A)^{-1} A^T$ se le conoce como la Penrose Pseudoinversa.
Pero que pasa si las columnas de $A$ no son linealmente independientes. Se sabe que si las columnas de $A$ son linealmente independientes $\mathcal{N}(A) = \{ 0 \}$.
Si las columnas de $A$ no son linealmente independientes $\mathcal{N}(A) \ne \{ 0 \}$ y no hay solucion unica.


Veamos un ejemplo simple donde las columnas de $A$ no son linealmente independientes.

**Ejemplo 1**:
Asuma el sistema lineal
$$2 x_1 + 3 x_2 = 5 \tag{0} $$.
Tiene un numero infinito de soluciones, por ejemplo $x_1=1, x_2=1$, otro $x_1=0$, $x_2=3/5$. Toda la linea que pasa por estos dos puntos es solucion del sistema $x_2 = (5 - 2 x_1)/3$.

Llevemos este problema a una forma matricial.
\begin{eqnarray}
A = (2, 3) \\
x = (x_1 , x_2)^T \\
b=5
\end{eqnarray}
El problema es $Ax = b$. Tiene un numero infinito de soluciones.
El minimo de $\| Ax - b \|$ existe, pero no es unico.

Si quisieramos usar minimos cuadrados (ecuaciones normales), tenemoms que encontrar $A^T A$.


\begin{eqnarray}
A^T A = \begin{pmatrix} 2 \\ 3 \end{pmatrix}
(2 \quad 3) =
\begin{pmatrix}
4  & 6 \\
6 & 9
\end{pmatrix}
\end{eqnarray}



Pero no ganamos mucho por que $A^T A$ no es invertible. de hecho $\det(A^T A) = 36-36 = 0$.

El espacio nulo de $A$ no es el trivial. De hecho por defnicion, el espacio nulo es los $(x_1, x_2)$ tales que

$$ 2x_1 + 3 x_2 = 0 \tag{1} $$.
Esta otra linea. Si $x_1=0$, $x_2=0$, $x_1=1$, $3x_2 = -2$, $x_2 = -2/3.$

Si tiene una solucion particular $x_p$ de $Ax=b$, entonces
$x=x_p + x_0$, donde $x_0 \in \mathcal{A}$, es solucion.

$$Ax = A(x_p + x_0) = A x_p + A x_0 = A x_p + 0 = A x_p = b  .$$
El espacio solucion es $x_p + x_0$ donde $x_0$ es cualquier punto del espacio nulo.

Como tengo un numero infinito de soluciones, escojo la mas pequena que es unica. Esta solucion mas pequena es perpendicular al espacio nulo (1)
El espacio perpendicular al espacio nulo esta dado por la ecuacion

$$-3 x_1 + 2 x_0 = 0 \tag{2}.$$

Tenemos el siguiente sistema: (Ecuaciones (0), (2)

\begin{eqnarray}
2 x_1 + 3 x_2 &=& 5 \\
-3 x_1 + 2 x_0 &=& 0
\end{eqnarray}
La solucion de este sistema es la solucion pseudoinversa.
Es la interseccion de la linea solucion con una perpendicular al espacio nulo (para hacerla la mas pequena).

La solucion del sistema es (multipliquemos la primera ecuacion por 3, la segunda por 2 y sumamos

$$ 13 x_2 = 5.$$
$$ x_2 = 5/13$$.

Reemplazamos en
$x_1 = (5 - 3 (5/13))/2 =10/13.$

La pseudosolucion es

\begin{eqnarray}
\begin{pmatrix}
x \\ y
\end{pmatrix}
=
\begin{pmatrix}
10/13 \\ 15/13
\end{pmatrix}
\end{eqnarray}
\end{eqnarray}
Veamos una grafica.

La matriz pseudoinversaa $A^+$, es tal que

$$ x = A^+ b . \tag{3}$$
Como hallamos la matriz $A^+$ ?
Escribamos el sistema (3)

\begin{eqnarray}
\begin{pmatrix}
10/13 \\ 15/13
\end{pmatrix}
=
\begin{pmatrix}
y_1 \\ y_2
\end{pmatrix}
5
\tag{4}
\end{eqnarray}

La pseudoinversa es
$$A^+ = \begin{pmatrix} y_1 \\ y_2 \end{pmatrix} .$$
Del sistema (4) encontramos que

$$A^+ = \begin{pmatrix} 10/65 \\ 15/65 \end{pmatrix} .$$



Compliquemos este problema un "pelito" mas.

**Ejemplo 2**:
\begin{eqnarray}
2 x_1 + 3 x_2 &=& 5  \quad , \quad \text{la original} \\
2 x_1 + 3 x_2 &=& 1  
\end{eqnarray}
Es obvio que este sistema no tiene solucion, por que esto implicaria $5=1$.

No hay solucion, que podemos hacer?
Para empezar vamos a formular el problema de forma matricial.
La matriz de $Ax=b$ es

\begin{eqnarray}
A =
\begin{pmatrix}
2 & 3 \\
2 & 3
\end{pmatrix}
\end{eqnarray}
Las columnas tampoco son linealmente independientes,
o sea que el espacio nulo no es el trivial. Pero anterior
no hay solucion.
Que pasa con $A^T A$,

\begin{eqnarray}
A^T A=
\begin{pmatrix}
2 & 2 \\
3 & 3
\end{pmatrix}
\begin{pmatrix}
2 & 3 \\
2 & 3
\end{pmatrix}
=
\begin{pmatrix}
8 & 12 \\
12 &18
\end{pmatrix}
\end{eqnarray}



vemos que $A^T A$ es el doble del $A^T A$ en el ejemplo anterior.
Como ejercicio dejamos como prueba de que si hay $n$ ecuaciones entonces $A^T A$ es $n$ veces el $A^T A$ de una sola ecuacion.

$A^T A$ tampoco tiene inversa (pues el espacio nulo no es trivial), las columnas de $A$ no son linealmente independientes.
Que hacemos?

Escribamos el sistema $Ax = b$

\begin{eqnarray}
\begin{pmatrix}
2 & 3 \\
2 & 3
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
=
\begin{pmatrix}
b_1 \\
b_2
\end{pmatrix}
\end{eqnarray}


Que este sistema tenga solucion es lo mismo que decir que $(b_1, b_2)^T$  pertenece al rango de $A$. Por que?
Por que este sistema se puede escribir como

\begin{eqnarray}
x_1
\begin{pmatrix}
2 \\ 2
\end{pmatrix} +
x_2
\begin{pmatrix}
3 \\ 3
\end{pmatrix}
=
\begin{pmatrix}
b_1 \\ b_2
\end{pmatrix}
\end{eqnarray}

La solucion, necesariamente es un multiplo del vector $(1,1)^T$.
Es espacio columna ($\mathbb{R}(A)$, es el generado por $(1,1)^T$
y la solucion **tiene** que estar en el espacio columna.

El vector $b = (b_1, b_2)^T = (1, 5)$.

Si no esta hay entonces **toca proyectarla**.
Ver notas a mano alzada.

El vector unitario en la direccion del rango es $(\sqrt{2}/2, \sqrt{2}/2)^T$. La proyeccion, como se explico en alguna clase anterior es:

\begin{eqnarray}
\left \langle
\binom{5}{1} ,
\binom{\sqrt{2}/2}{\sqrt{2}/2}
  \right \rangle
\binom{\sqrt{2}/2}{\sqrt{2}/2}
= \binom{3}{3}
\end{eqnarray}



Veamos que esta es la mejor solucion en el sentido de minimos cuadrados. Sea $b_0$ el vector "distancia" entre $b$ y $b_r$,
es decir $b_0=b - b_r$. Tenemos un triangulo rectangulo con lados $b, b_r, b_0$, por el teoerma de Pitagoras


$$ \| b \|^2 = \| b_r \|^2 + \| b_0 \|^2 .$$
Vamos a ver que $\min_x \| Ax - b \| = \| b_0 \|. $



image.png

Veamos (analiticamente)
\begin{eqnarray}
\| Ax - b \|^2 &=& \langle Ax - b, Ax -b \rangle \\
&=& \langle Ax - b_r - b_0, Ax - b_r - b_0 \rangle \\
&=& \langle (Ax - b_r) - b_0 ,  (Ax - b_r) - b_0 \rangle \\
&=& \| Ax - b_r \|^2 - 2 \langle Ax - b_r, b_0 \rangle + \| b_0 \|^2 \\
\end{eqnarray}
Pero $\langle Ax - b_r, b_0 \rangle = 0$. Por que? pues
$Ax -b_r \in \mathcal{R}(A)$.  y $b_0 \perp \mathcal{R}(A)$.
entonces

\begin{eqnarray}
\| Ax - b \|^2 = \| Ax - b_r \|^2 + \| b_0 \|^2.
\end{eqnarray}
El primer termino puede convertirse en 0 por que $b_r \in \mathcal{R}(A)$ pero el segundo no cambia. El minimo de todos esos valores es el segundo termino $\| b_0 \|^2$.
Esto resuelve el problema de minimos cuadrados.

De acuerdo a la solcion el punto de proyeccion es $(3,3)^T$
y las ecuaciones se  cambian.
\begin{eqnarray}
2 x_1 + 3 x_2 &=& 3  \\
2 x_1 + 3 x_2 &=& 3  
\end{eqnarray}
El espacio solucion es la solucion de la ecuacion $2 x_1 + 3 x_2=3$ que es una linea recta. (la linea azul en la grafica de abajo).


El espacio solucion al nuevo sistema $A x = b_r$ es entonces $\mathcal{R}(A)$.  Pero hay una infinidad de soluciones.
El espacio nulo es $2 x_1 + 3 x_2 = 0$, y el normal al espacio nulo es $-3 x_1 + 2 x_2 = 0$.

Encontramos las dos ecuaciones

\begin{eqnarray}
2 x_1 + 3 x_2 = 3 \\
- 3 x_1 + 2 x_2 = 0
\end{eqnarray}
De la misma forma por reduccion Gaussiana

\begin{eqnarray}
x_1 \approx 0.69231 \\
x_2 \approx 0.4514
\end{eqnarray}
Construya con esto $A^+$.
