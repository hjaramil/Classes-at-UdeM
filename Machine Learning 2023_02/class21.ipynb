# Multisets (Multi-conjuntos)
Un multiset (multiconjunto) es un  conjunto donde se pueden repetir elementos. Por ejemplo

$$ M = \{a, a, a, b, b, c \} .$$
En que casos ocurren cosas de estas. En muchos y es el objeto de los proximos 30 minutos. Por ejemplo piense en la factorizacion prima de un numero,

$$ 360 = 2^3 3^2 5^1 .$$
Esto como multiconjunto se puede escribir como $\{2,2,2, 3,3, 5 \}$.

Otro ejemplo, la palabra "bookkeeper" tiene letras repetidas.
Podemos hacer una tabla donde se muestre la frecuencia de cada letra.

| Caracter | multiplicidad |
| ---- | ----- |
e | 3
0 | 2
k | 2
b | 1
p | 1
r | 1

El multiconjunto es $\{e, e, e, o, o, k, k, b, p, r \}$.
El multiconjunto lo podemos formalizar como una funcion de un conjunto $A$, a los numeros naturales

\begin{eqnarray}
m : A &\to& \mathbb{N} \\
    x &\mapsto& m(x) = \text{multiplicidad de } x
\end{eqnarray}
En el caso de "bookkeeper", $A=\{b,o,k,e,p,r \}$.

Vamos a contar multiconjuntos.

## El orden importa (permutaciones).
Cuantas palabras distintas podemos pormar con el multiconjunto "bookkeeper". Podriamos ingenuamente pensar que como
$n=10$ es el numero de letras de la palabra, entonce $10!$ es la solucion al problema. Sin embargo ojo!!!,  si intercambia
la segunda "o" con la tercera "o" de la palabra no pasa nada.
Ah bueno, como hay 2 "o" dividimos por 2!, Pero ojo!!!, hay 3 "e",
entonces dividimos por $3!$, entonces realmente el conteo de **todas las palabras** que se pueden crear a partir de las letras de la palabra "bookkeeper" es:


$$ \text{conteo} = \frac{10!}{2! 2! 3! 1! 1! 1!} .$$

Digamos que $n$ es el cardinal de todos los elementos (repetidos o no) del mutiset $(A,m) = \{e,e,e,o,o,k,k,b,p,r \}$. Es decir

$$ n= \#(A,m) = m(e) + m(o) + m(k) + m(b) + +m(p)+ m(r) 3 + 2+ 2+ 1 + 1 + 1 = 10.$$






La formula general, la podemos abstraer del analisis anterior.


$$\text{conteo} = \frac{n!}{m_1 !  m_2! \cdots m_k !} .$$

Relacionemos este conteo con un problema interesante del algebra.

### Multinomial:
Piense en la siguiente formula.

$$(x_1 + x_2 + \cdots + x_k)^n =\sum_{i_1 + i_2 + \cdots i_k=n}
c_{i_1 i_2 \cdots i_k} x_1^{i_1} x_2^{i_2} \cdots x_{k}^{i_k} .$$

Dos preguntas:
* Quien es $c_{i_1 \cdots i_k}$?
* Cuantos elementos satisfacen $i_1 + i_2 \cdots + i_k=n$ (Diofantina).

Pregunta 1:
Piense que tenemos un multiset donde

$$ A = \{ x_1, x_2, \cdots , x_k \} \quad \text{y} \quad m_j= m(x_j)= i_j.$$
Entonces
Por la ley conmutativa, no importa el orden.

$$ c_{i_1 i_2 \cdots i_k} = \frac{n!}{i_1! i_2! \cdots i_k!} = \binom{n}{i_1 i_2 \cdots i_k} .$$

Lo primero que podemos decir que es esta es una generalizacion de la binomial. Tome el problema

$$ (x_1 + x_2)^n =  \sum_{i=1}^n \binom{n}{k} x_1^i x_2^{n-i} .$$

En este caso, tenemos solo dos grupos (los de $x_1$ y los de $x_2$. Si un grupo toma $k$, el otro toma $n-k$ y la multinomial arroja


$$\frac{n}{n! (n-k)!} = \binom{n}{k} .$$

Esta binomial sirve para millones de cosas, pero veamos una aplicacion bonita de la teoria de conjuntos.
La binomial cuenta el numero de subconjuntos de $k$ elementos de un conjunto de $n$,  Tomemos el binomio

$$ (x_1 + x_2)^n =  \sum_{i=1}^n \binom{n}{k} x_1^i x_2^{n-i} .$$
Escojamos $x_1=x_2=1$, entonces

$$ 2^n =  \sum_{i=1}^n \binom{n}{k}  .$$
El resultado muestra que el numero total de subconjuntos de un conjunto con $n$ elementos es $2^n$.

Pero veamos como podemos probar esto usano la propiedad de conteo de las cajas. Ver notas a mano alzada.





## El orden no importa (combinaciones).
Este problema se puede interpretar de varias formas.

1. El numero de soluciones $(x_1, x_2, \cdots, x_k)$ de la ecuacion diofantina $x_1 + x_2 + \cdots + x_k=n$.
Como la suma es conmutativa, el orden no importa. Ahora
los $x_i$ pueden ser repetidos o no.
$x_1 + x_2 + x_3 = 4$,
2. El numero de combinaciones de $k$ objetos tomados $n$ a la vez
con repeticion.
$(2,2,0), (1,1,2), $
3. El numero en que $n$ objetos identicos pueden ser distribuidos en $k$ cajas.

Para resolver este problema recurrimos al metodo de
[stars and bars](https://en.wikipedia.org/wiki/Stars_and_bars_(combinatorics)#:~:text=In%20the%20context%20of%20combinatorial,his%20classic%20book%20on%20probability.)

En vez de estrellas yo uso el signo "+".  Voy a explicar este metodo con un ejemplo. De cuantas formas podemos sumar
$x_1 + x_2 + x_2 = 2$.

Aca estan las formas

\begin{eqnarray}
1 + 1 + 0 &=& 2
1 + 0 + 0 &=& 2
0 + 1 + 1 &=& 2
2 + 0 + 0 &=& 2
0 + 2 + 0 &=& 2
0 + 0 + 2 &=& 2
\end{eqnarray}
Al numero '1' lo reemplazo con una barra "|", al "+" lo dejo quieto, al 0, el "2" es "||" lo ignoro

\begin{eqnarray}
|+|+  \\
|++| \\
+|+|  \\
||++ \\
+||+ \\
++||
\end{eqnarray}
Tenemos solo dos simbolos $+$ y $|$. De cuantas formas
puedo colocar estos dos simbolos en un expresion de 4 simbolos.

Este produce

$$\text{conteo} = \binom{4}{2} = \frac{4!}{2! 2!} = \frac{4 \times 3 \times 2!}{2! 2!} = 6.$$

Generalicemos. En general en la Ecuacion
$$x_1 + x_2 + \cdots + x_k = n  .$$

vamos a encontrar expresiones de la forma

$$ ||+|||+ \cdots +||+ | .$$
En total, el numero de barras es $n$, y el numero de $+$ es
$k-1$.

Piense en el problema, cuantas palabras puede formar con $n+k-1$ simbolos, donde hay $n$ repetidos y $k-1$ repetidos. Multiset (recuerde el ejemplo de "bookkeeper"

$$\text{conteo} = \frac{(n+k-1)!}{n! (k-1)!}= \binom{n+k-1}{n} = \binom{n+k-1}{k-1} .$$
Este resuelve el numero de soluciones a la ecuacion
$x_1 + x_2 + \cdots x_k=n$.

En la literatura es comun ver que $n$ y $k$ se intercambian.
Es decir

Este resuelve el numero de soluciones a la ecuacion
$x_1 + x_2 + \cdots x_n=k$.
y obtenemos

$$ \begin{pmatrix} \binom{n}{k} \end{pmatrix} = \binom{n+k-1}{k} = \frac{(n+k-1)!}{k! (n-1)!} = \frac{n(n+1)(n+2) \cdots (n+k-1)}{k!} = \frac{n^{\overline{k}}}{k!}  .$$
donde $n^{\overline{k}}$ es el
[raising factorial](https://en.wikipedia.org/wiki/Falling_and_rising_factorials)

Observen la simetria.

* $$\text{choose} = \binom{n}{k} = \frac{n^{\underline{k}}}{k!} $$
* $$\text{multichoose} = \left (\binom{n}{k} \right ) = \frac{n^{\overline{k}}}{k!} $$
