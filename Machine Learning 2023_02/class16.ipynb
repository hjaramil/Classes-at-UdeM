Desventajas del "overfitting"

* Cuando entrena el sistema, se crean soluciones que siguen el ruido con detalle.

* El ajuste exagerado puede crear altas oscilaciones y varianza en la solucion. Hay un tema que estudiamos mas adelante en este curso llamado "bias versus variance".

* El computo para funciones sobreajustadas es mucho mayor.

* Por ultimo, la solucion que se obtiene esta lejos de ser buena para probar con **datos nuevos**.

## Underfitting:
La solucion obtenida por subajuste no captura los detalles que necesita el modelo. De cierta forma "underfitting" es el opuesto a "overfitting".

Desventajas de underfitting.

* extrema simplicidad del modelo que no se ajusta a los datos.
* no oscila lo sufificente.
* inaceptable en la prediccion con nuevos datos.

### Como controlar el overfitting.

* Reducir el numero de "features" (atributos)
* Aplicar regularizacion (enseguida muestro ejemplos)
* usar mas datos para mejorar el modelo. Separar los datos en entrenamiento y prueba. O tres grupos, training, cross-validation, training. De esto hablamos mas adelante.
Hay otro metodo, que en minutos describimos. Se llama $k$-fold cross-validation.
* usando tecnicas donde se tiene datos para entrenar (training data) y datos para prueba (testing data). Con esto se diagnostica si hay overfitting.

## K-fold cross validation.
Tenemos un grupo de pruebas para entrenar un sistema de ML.
Habiamos dicho que no usar todos los datos para entrenar podria salvarnos del flajelo de "overfitting". En que consiste el k-fold cross-validation?

Los datos ($m$ )se dividen en $k$ "cajitas".

* Usamos las cajitas $2,3, \cdots , k$ para entrenar el sistema.
y la cajita 1 para probar.
* Usamos las cajitas $1, 3, 4, \cdots, k$ para entrenar y la cajita $2$ para probar.
* Asi suscesivamente, usamos las cajitas
$1,2, \cdots, p-1, p+1, \cdots, k$ para entrenar y la cajita $p$ para probar.


En cada caso calculamos el error con alguna metrica de las que hemos estudiado y vamos a estudiar en este curso.
Se promedian todos estos errores y con esto diagnosticamos el algoritmo. Si hay overtting o no.

### Formulacion del problema.
Es comun que un problema se reduzca a solo dos estados.
Falso y verdadero, 1 o 0, blanco o negro. A este variables se les llama **binarias**.  La pregunta es si podemos usar lo que ya sabemos de regresion lineal para este problema. La regresion lineal funciona bien para variables continuas, pero esta es una variable discreta. Seria bueno tener una hipotesis de la forma


$$h(\Theta) = X \Theta = Y .$$
donde $X$ son los datos, $Y$ son las etiquetas (son binarias)
y $\Theta$ son los parametros a **aprender**.

Si quisieramos seguir en esta tarea, pensariamos en la minimzacion una funcion objetivo de la forma

$$ \min_{\Theta} = \min_{\Theta} \frac12 \| X \Theta - Y \|^2 .$$
El problema es que la variable $Y$ es discreta y solo toma dos valores. Como resolver este problema?
