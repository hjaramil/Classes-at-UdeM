Trabajamos sobre los reales $\mathbb{R}$ y complejos $\mathbb{C}$.

**Propiedades del producto interno**:
Asumamos $\alpha \in \mathbb{C}$.

* El producto interno sobre los reales es conmutativo, sobre los complejos no. Encontrar un ejemplo.

* Extraccion del escalar:

\begin{eqnarray}
\langle \alpha u, v \rangle &=& \overline{\alpha} \langle u , v \rangle  \\
\langle u, \alpha v \rangle &=& \alpha \langle u , v \rangle .
\end{eqnarray}
Se prueba de la definicion y las propiedes de matrices.

* El producto interno es distributivo.

\begin{eqnarray}
\langle u, v \pm w \rangle &=& \langle u , v \rangle \pm \langle u, v \rangle \\
\langle u \pm v , w \rangle &=& \langle u, w \rangle \pm \langle v, w \rangle.
\end{eqnarray}


Definicion **matriz autoadjunta**: La matriz $A$ se dice que es autoadjunta (self adjoint, Hermitian) si
$A^* = A$.  


Definicion: **norma**, **vector unitario**. Asumamos un vector $v \in \mathbb{C}^n$. Definimos la norma del vector como $\| v \| = \sqrt{ \langle v, v \rangle} $.
El ejemplo mas simple: $v=(a,b)^T$, $a,b \in \mathbb{R}$. Entonces verifique que $\| v \| = \sqrt{a^2 + b^2}$.
Y la norma es la metrica Euclidiana o Pitagorica.
La definicion de norma es mucho mas general en el area de analisis funcional pero aca nos quedamos con esta.
Un vector $u$ es unitario si $\| u \| = 1$.

A la norma como la definimos arriba se le llama $\ell_2$. O la norma 2. Por eso a veces se usa la notacion $\| v \|_2$.

Definicion **norma Frobenius**: La norma Frobenius de una matriz $A \in \mathbb{C}^{m \times n}$, se define como

$$ \| A \|_F = \sqrt{ \sum_{i=1}^m \sum_{j=1}^n | a_{ij} |^2 } $$

Vamos a asumir el teorema de Pitagoras:
Asumamos $x,  y \in \mathbb{R}^n$.

$$ \| x \|^2 + \| y \|^2 = \| x + y \|^2 $$


Mostremos la ortogonalidad usando el teorema de Pitagoras:

\begin{eqnarray}
 \| x + y \|^2  &=& \langle x + y, x+y \rangle \\
 &=& \langle x, x \rangle + \langle x, y \rangle + \langle y, x \rangle + \langle y , y \rangle \quad \text{propiedad distributiva del producto interno}\\
 &=& \langle x , x \rangle + 2 \langle x, y \rangle + \langle y, y \rangle \\
 &=& \| x \|^2 + 2 \langle x, y \rangle + \| y \|^2.
\end{eqnarray}
Compare esto con el teorema de Pitagoras, y eso cierto si y solo si

$$ 2 \langle x, y \rangle = 0$$
O sea que $\langle x, y \rangle = 0$ y los vectores son ortogonales.

Definicion **Matriz identidad**: La matriz $n \times n$ $I$ se define por la regla del **Delta de Kroneckere**, es decir

\begin{eqnarray}
\delta_{ij} = \left \{
    \begin{array}{cc}
    1 & \text{si   }  i = j \\
    0 & \text{si   } i \ne j
    \end{array}
    \right .
\end{eqnarray}


En este caso la matriz identidad se define como


\begin{eqnarray}
I =
\begin{pmatrix}
1 & 0 & 0 & \cdots & 0 \\
0 & 1 & 0 & \cdots & \vdots  \\
0 & 0 & 1 & \cdots & \vdots  \\
\vdots & \vdots & \vdots & \ddots & 0 \\
0 & \cdots & \cdots & 0 & 1
\end{pmatrix}
\end{eqnarray}

Definicion **inversa**: Sea $A$ una matriz $n \times n$. Si existe $B$ tal que $AB = BA = I$, decimos que $B$ es la inversa de $A$ y escribimos $B= A^{-1}$. Si existe la inversa de $A$, decimos que $A$ es **no singular**, de lo contrario decimos que $A$ es **singular**.

Una propiedad importante de la identidad es que $AI = A=IA$.
Teorema: **inversa y el determinante**:

* Si $A,B$ son matrices $n \times n$, entonces

$$ \det(AB) = \det(A) \det(B).$$

* Si $\det(A) \ne 0$, entonces la inversa de $A$ existe.


Teorema: Si $AB=I$, y $A,B$ son cuadradas, entonces $BA=I$,
es decir $AB=BA=I$.


Prueba. Sea $AB=I$.

$$ B = B I = B (AB) = (BA) B  \tag{1} $$
Como el $\det I = 1$ y de $\det(AB) = \det (A) \det(B)$ entonces
$\det(A) \det (B)=1$. De forma que $\det(B) \ne 0$, y por la tanto existe $B^{-1}$.

Multiplicamos la ecuacion (1) por $B^{-1}$ a ambos lados a derecha


$$ B B^{-1} = (BA) B B^{-1}  $$
Es decir

$$ I = BA $$
De esta forma si $A^*A = I$ entonces $A A^*=I$ . Otra forma de decirlo es que las matrices **ortogonales** tiene la propiedad de que la adjunta es la inversa. Es decir, si $A$ es ortogonal.

$$ A^{-1} = A^*$$

Recuerden que si $A$ es real $A^*  = A^T$ y si es ortogonal

$$ A^{-1} = A^T$$.

Otra propiedad interesante de las matrices ortogonales es que son operadores **isometricos**. Es decir, no cambian el tamanho del objeto sobre los que operan. Mas concretamente.


Teorema: Si $A$ es ortogonal $x \in \mathbb{C}^n$, entonces
$\| A x \| = \| x \|$.

Prueba:

$$\| A x \|^2 = \langle Ax, Ax \rangle = (A x)^* Ax = x^* A^* A x = x^* I x = x^* x = \langle x, x \rangle = \| x \|^2 $$
y sacando la raiz cuadrada

$$ \| Ax \| = \| x \|. $$

Teorema (**proyeccion**). La proyeccion de un vector $v$ a una linea $L$ con direccion $u$ esta dada por

$$ b = \overline{\langle v, u \rangle}  u$$
La norma de la proyeccion esta dada por

$$ \|b \| = | \langle v, u \rangle |.$$
puesto que $u$ unitario.

Veamos la prueba: (nos basamos en la grafica).

$$v = b + d = \alpha u + d. $$
Tomamos el product interno con $u$ para encontrar

$$\langle v, u \rangle = \langle \alpha u, u \rangle  + \langle d, u \rangle = \overline{\alpha} \langle u, u \rangle + \langle d, u \rangle. $$
Recuerden que $u$ es unitario, o sea $\langle u, u \rangle = 1$.
Ademas recuerde que $\langle d, u \rangle = 0$ (son ortogonales).
Encontramos entonces:


$$ \langle v, u \rangle = \overline{\alpha}$$
y tomando el conjugado

$$ \alpha = \overline{\langle v, u \rangle} $$.

De forma que la proyeccion es

$$ b = \alpha u = \overline{ \langle v, u \rangle } u.$$

Ultima definicion de hoy y es **espacio rango (columna) y espacio nulo**


Definicion: **espacio rango y nulo**: Dada una matriz $A$, $m \times n$, el espacio rango (columna) es el espacio generado por las columnas de $A$. La notacion es $\mathcal{R}(A)$. Es decir,
si $A$ tiene columnas $A_i$, el espacio rango es la combinacion lineal de todas estas columnas, es decir, los vectores de la forma
$\sum_{i=1}^n \alpha_i A_i$, con $\alpha \in \mathbb{C}$.
El espacio **nulo** (el **kernel**), en ingles **null** es el conjunto de todos aquellos $x$, tales que $Ax=0$.
