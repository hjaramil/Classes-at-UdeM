## AUC-ROC
AUC= Area under the curve.
Cual curve? la ROC:

[Receiver Operation Characteristic](https://en.wikipedia.org/wiki/Receiver_operating_characteristic)

Esta curva es una parametrica donde el parametro es el umbral $\gamma$ que se usa para separar una clase de otra. Falsos contra verdaderos, blancos contra negros, etc.
Como se calcula?
El eje $x$ de esta curva es el FPR.
 $$\text{FPR} =  \frac{\text{FP}}{\text{FP + TN}}(\gamma) .$$
El eje $y$ es el True positive rate.
 $$\text{TPR} =  \frac{\text{TP}}{\text{TP + FN}}(\gamma) .$$

 Esto lo construyeron por primera vez durante la segunda guerra mundial para estudiar senhales de radar. Aca las dos clases son senhal vs ruido.
 Esta curva es como un infinidad de matrices de confusion con respecto al parametro umbral $\gamma$

 Con varios ejemplos grafico explico como se genera esta curva.







image.png

Asumamos que el umbral $\gamma=0$ (no esta en la figura.
Entonces todos los positivos estan por encima de 0, o sea que
TP=4, FP=4, TN=0, FN=0. En este caso

$$ TPR = \frac{TP}{TP + FN}= \frac{4}{4} = 1.$$
$$ FPR = \frac{FP}{FP + TN}= \frac{4}{4} = 1.$$

El otro extremo es cuando $\gamma = 1$.
En este caso
TP=0, TN=4, FP=0, FN=4.

$$ TPR = \frac{TP}{TP + FN}= \frac{0}{4} = 0.$$
$$ FPR = \frac{FP}{FP + TN}= \frac{0}{4} = 0.$$


Tenemos calculados en la  curva ROC, 6 puntos.

$$(1,1), (0.25,1), (0.25, 0.75), (0, 0.75), (0, 0.5), (0,0) .$$

Se grafican esos puntos y esta es la ROC

image.png

En cuanto a la interpretacion de esta curva.

* En la diagonal FPR = TPR, no es muy bueno.
* Arriba de la diagonal es bueno. Entre mas arriba mejor.
* Debajo de la diagonal es malo. Entrer mas abajo peor.
* El area bajo esta curva se llama AUC. Entre mas grande mejor.

En el caso de la figura, el area (verde), AUC es
$$ 1 - (0.25)^2 = 0.9375.$$

Que pasa cuando el numero de clases es mayor que 2?

Como es la matriz de confusion?
Para este caso la matriz de confusion es una matriz de orden $K \times K$. A lo largo de la cada fila (horizontal) etiquetamos los valores reales y a largo de cada columna etiquetamos los valores predichos. Solo aquellos valores de la diagonal corresponden a valores en los cuales la prediccion corresponde con la realidad. Es decir TP o TN

Accuracy (exactitud) es el numero de elementos en la diagonal, dividido por el total de muestras $m$.

Es decir

$$\text{accuracy} = \frac{\sum_{\ell=1}^m \ell_{ii}}{m} .$$


# Analisis combinatorio (combinatoria)

Usamos un reglas basicas.

* La regla de la multiplicacion.
Si tenemos $k$ cajas con elementos $b_k$ en cada caja.
Cuantas combinaciones podemos tener?

El total de combinaciones es

$$\text{conteo} = b_1 \times b_2 \times \cdots \times b_k .$$
Si todas las cajas tienen el mismo numero de elementos $n$, tenemos entonces que el tota es $n^k$.

$$\text{conteo} = n^k .$$

image.png

Algunos ejemplos donde esto es util.
* $k$-tuplas. Cuantas listas como vectores $(x_1, x_2, \cdots, x_k) \in \mathbb{R}^k $ existen cuando los $x_i$ toman un rango de valores $b_i$.

* Cuente numeros de varios digitos donde cada digito puede tomar varios valores. Por ejemplo los numeros de 3 digitos cuando cada digito toma los valores $0, 1, \cdots, 9$, son $1000=10^3$.
Si no puede usar el 9 entonce seria $9^3=243$.

* Cuente el numero de placas de carro distintas con 2 letras y tres numeros.

$$ 27 \times 27 \times 10 \times 10 \times 10 =729000 .$$

##  Conjuntos (Sets)
Un conjunto (informalmente) es una lista no ordena con objetos.
No hay repetidos

* no hay repetidos
* el orden no importa.

Asumamos el modelo de $k$ cajas, podemos asignar a cada caja uno de $n$ ($n \ge k$) elementos, pero **ojo** no puede haber repeticion. O sea que si el elemnto "a" se asigno a la primera caja, ya no se puede asignar a ninguna otra. (is taken).

## Permutaciones

$$\text{conteo} = n \times (n-1) \times \cdots \times (n-k+1) = \frac{n!}{(n-k)!} = n^\underline{k} .$$
(ver notas a mano alzada).
En este experimento, el orden esta establecido. Usted toma uno,
para la caja uno, luego otro para la caja 2 y asi......
Pero por ejemplo, el conteo, sobre el conjunto $\{a, b, c \}$,
tomando $k=2$,

La primera caja puede tender $a,b,c$ (tres posibilidades), pero
si es $a$, la segunda, solo tiene $b,c$, si esta es $b$ la tercera solo tienes $c$. Pero ojo, en la primera pudimos comenzar con $b$ en vez de $a$. Estamos contando todas las posiblidades y importando el orden. No se lo mismo que comience con $a$ o con $b$ o con $c$. El orden importa.

##  combinaciones.
Que tal si el orden no importa. Es decir, por ejemplo $(ab)=(ba)$.
En el ejemplo anterior $(ab) \ne (ba)$.

Cuando el orden no importa se llaman combinaciones, cuando el orden importa se llama permutaciones.

Vuelva al modelo de la caja. Cada elemento, vamos a ordenarlo

$(a_1 a_2 \cdots a_k) $ este elemento se conto muchas veces por que por ejemplo
$(a_k \cdots a_2 a_1)$ se conto como otro. Cuantas veces se conto?
Se conto $k!$.

Para poder aseguar que el orden no importa y que, por ejemplo, $(ab)=(ba)$ entonces necesitamos dividir por $k!$ la expresion anterior, y obtenemos

$$\text{conteo} = \frac{n!}{(n-k)! k!} = \binom{n}{k} = \frac{n^{\underline{k}}}{k!} .$$
El ejemplo clasico de esto es el numero de subconjuntos de un conjunto. Esto lo vemos en la proxima clase.


Resumen: Hay cuatro posibilidades para los experimentos con conjuntos.
Piense en una bolsa con $n$ bolas y extraemos una por una $k$ bolas.

1. Importa el orden  (permutaciones)
    * Con reemplazo (repeticion) $n^k$
    * Sin reemplazo (repeticion) $n^{\underline{k}}= n (n-1) (n-2) \cdots (n-k+1)$
2. No importa el orden (combinaciones)
    * Con reemplazo (repeticion) $n^k/k!$
    * Sin reemplazo (repeticion) $n^{\underline{k}}/k! $
