## Machine Learning (Introduccion)
Los topicos que nosotros vamos a discutir en ML, son

* Prediccion
* Regresion
* Clasificacion
* Clustering
* Problemas de control (no los vemos)
* Generacion (por ejemplo GAN: Generative Adversary Netwoks).

La programacion clasica usa datos y procedimientos para obtener resultados. El machine learning usa datos y resultados (respuestas) para obtener procedimientos que resuelven problemas.
INcluso puede no tener resultados y con los solos datos resolver problemas. Por ejemplo agrupacion.

Programacion clasica:
 * Datos a=5, b=4
 * procedimientos a+b
 * resultados 9

Machine learning.
* Datos: 100 cartas de figuras de animales
* resultados: etiquetas (labels), los nombres de los animales.
* procedimiento. Un ninho es entrenado por el padre para reconocer los animales en las tarjetas. Pero se le dan el animal y su nombre (etiqueta). Es decir es **supervisado**.
Una vez entrenado, el ninho esta listo para recibir cartas de animales y  adivinar su nombre.




El machine learning se conforma por varios topicos.


* Aprendizaje Supervisado: Se usan etiquetas. Es decir se tiene las respuestas. Por ejemplo

$Ax = b$, regresion lineal. $b$ es un vector con las etiquetas (las respuestas). La programacion clasica, te da $A$, $x$ para calcular $b$. La regresion lineal (que es parte de ML) te da $b$, y $A$ para que calcules $x$. En este sentido es un probema de inversion. $x=A^{-1}b$,  pero $A$ no necesariamente tiene inversa. Este problema lo vamos a analizar con detalle en el siguiente capitulo.
* Aprendizaje No supervisado: En este caso no se dan las etiquetas. Por ejemplo, al ninho se le dan 100 tarjetas con figuras de animales. En ningun momento se le esta dando la informacion de que es un elefante, o jirafa o etc. El ninho
simplemente toma las cartas y las parte en grupos. Grupos de
elefantes, de jirafas, de cebras, etc. A este tipo de procedimientos, se les llama **clustering**.
* Aprendizaje reforzado (reinforcement learning): Este se usa para los juegos. Existe un **agente** (el jugador) y un **ambiente**. El agente lucha contra el ambiente y si pierde, lo tiene en cuenta, si gana tambien. Si pierde tiene un penalti, si gana una recompenza. Al comienzo el jugador pierde todo el tiempo, por que no tiene idea, pero a medida que va perdiendo va aprendiendo y eventualmente comienza a ganar y volver experto.
En este curso no estudiamos reinforcement learning.

El curso trata de los primeros.
Incluso existe otra parte llamada **aprendizaje semi-supervisado**.
Puede ser que se tengan algunas etiquetas y otras no.

Ejemplo. Sistemas de recomendacion (Netflix). Usted ve una pelicula, la califica si o no. Si "si" esta dejando una etiqueta, si "no" no la esta etiquetando. Netflix tiene millones de usuarios, (digo yo) que menos del 10% califica las peliculas.
Si usted quiere que Netflix le haga buenas sugerencias, podria calificarlo.

Ejemplo. YouTube. El sistema aprende aunque usted ni siquiera calfique las canciones o videos. Por que? por que internamente esta almacenando toda esa informacion como el numero de veces que ve un video, o el tipo de videos que ve y con que frecuencia.
