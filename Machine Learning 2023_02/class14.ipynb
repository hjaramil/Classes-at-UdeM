# Metodo de "steepest descent" (Descenso maximo)
Consideramos la funcion de costo $J$ cuadratica (correspondiente a un problema lineal $X \theta =y$).
$$ J =  \frac12 \| X \theta - y \|^2 =\frac12 \| h(\theta, x) - y \|^2 = \frac12 \sum_{i=1}^m ( h(\theta, x^{(i)}) - y_i)^2 .$$
donde $h(\theta, x^{(i)}) = x^{(i)} \theta$, con $x^{(i)}$ la fila $i$
de la matriz $X$.

Comenzamos con calcular el gradiente de $J$. Es decir,


Hallemos el gradiente.

$$\frac{\partial J}{\partial \theta_k} = \sum_{i=1}^m (h(\theta, x^{(i)}) - y_i) x_{ik}  \tag{1} .$$
donde $x_{ik}$ es la coordenada $k$ del vector $x^{(i)}$.

Si la funcion es convexa tiene minimo. Una funcion es convexa en un domionio $D$ si el segemento (o plano, o hiperplano) que une dos partes de la grafica esta totalmente sobre la grafica.
Para funciones reales $f: A \subset \mathbb{R} \to \mathbb{R}$, una
funcion es convexa si $f''(x) \ge 0$, $\forall x \in A$
For exemplo $f(x)=x^2$, $f''(x)=2 > 0 $.



La ecuacion (1) en forma vectorial es:


$$\nabla J(\theta) = X^T( X \theta - y) .$$
De hecho si $\nabla J = 0$, encontramos por n-esima vez

$$ X^T X \theta = X^T y. \tag{2} $$

Recordemos que para descender queremos iterar
$$\theta^{(i)} = \theta^{(i-1)} - \alpha \nabla f(\theta^{(i-1)}) .$$
Aca hay dos preguntas fundamentales.

1. El gradiente $\nabla f(\theta)$ apunta al minimo?
2. Que tan grande o pequenho debe ser $\alpha$ para bajar los mas rapido posible.

1. No siempre.  Casi nunca. Excepto en una esfera. O en una elipse si se para en los vertices.
De esta pregunta non discutimos mucho.

De la segunda pregunta, vamos a discutir ahora.
Estudiamos el "search parameter". A este tipo de busqueda se le llama **line search**.

En la ecuacion (2) por simplificacion llaemos
$A = X^T X$, $b=X^T y$. Entonces estamos resolviendo el problema
$$ A \theta = b.$$
con $A$ definida (si no entonces $A=(X^T X + \lambda I), \lambda>0$)

$$J(\theta) = A \theta - b .$$
Estamos parados en $\theta_0$ y queremos a vanzar hacia
$\theta_1 = \theta_0 - \alpha \nabla J(\theta_0)$.
Tenemos el siguiente problema, cual es $\alpha$ para que

$$ \varphi(\alpha) = J[ \theta_0 - \alpha \nabla J(\theta_0)].$$
sea minimo.










$\varphi$ es una funcion $\mathbb{R} \to \mathbb{R}$.
El resultado se hallar tomando la derivada e igualandola a 0.
Usamos la regla de la cadena en funciones de varias variables.

$$ \varphi'(\alpha) =  \nabla J[\theta_0 - \alpha \nabla J(\theta_0)] \cdot \nabla J(\theta_0) = 0  \tag{4} .$$
Por simplificacion definamos

$$\theta_1 = \theta_0 - \alpha \nabla J(\theta_0) .$$

Multipliquemos esta ecuacion por $A$ (izquierda) ambos lados
y restemos $b$.

$$ A \theta_1- b = A \theta_0 -b - \alpha A \nabla J(\theta_0)  .\tag{3} $$
Vamos a llamar (simplificando) el residual

$$ r_0 = A \theta_0 - b = \nabla J(\theta_0) .$$
$$ r_1 = A \theta_1 - b = \nabla J(\theta_1) .$$
Escribamos (3) con esta nueva notacion


$$ r_1 = r_0  - \alpha A \nabla J(\theta_0)  \tag{5} .$$
Note que $r_1 . r_0 = 0$. (por la ecuacion (4))

Vamos a usar esto para elimitar $r_1$ de la ecuacion de (5).

Es decir, pre-multiplicamos, a izquierda, con producto punto, la ecuacion (5)
por $r_0$, entonces

$$ r_1 \cdot r_0 = r_0 \cdot r_0  - \alpha r_0^T A r_0   .$$
es decir
$$ 0 = \| r_0 \|^2 - \alpha r_0^T A r_0 .$$
y

$$ \alpha = \frac{\| r_0 \|^2}{r_0^T A r_0} .$$

Pregunto yo: Habra algo mas efectivo que esto?

La respuesta es si. En 1952, tres personas distintas resolvieron este problema para converger en solo 2 iteraciones en $\mathbb{R}^2$ y en $n$ iteraciones en $\mathbb{R}^n$.  Las tres personas fueron, Magnus Hestenes, Eduard, Stiefel y Cornelius Lanczos.
Hestenes y Stiefel publicaron el articulo (lo pueden bajar de Wikipedia) en 1952.

Digo yo: "si su sistema es lineal y jigantesco $A \theta  = b$, el mejor metodo para resolverlo es CG.

Pero desafortunadamente los sistemas lineales son pocos. Debemos
buscar otras alternativas.

Hay muchas para funciones altamente complicadas con gradiente no lineal.
* Metodos de Newton: poco se usan por costosos. Necesitan el calculo de la Hessiana y este es muy costoso para sistemas enormes.
* Metodos de Quasi-Newton. Estos se usan mucho. El mas popular es
el L-BFGS. La L es por limited memory, y las letras son por
Broyden-Fletcher-Goldfrab-Shanno.  Es como un Newton pero la Hessiana la simplican a una diagonal.

En Deep learning se usan aceleradores de gradiente, que no vemos aca.
* Momento
* Nesterov
* AdaGrad
* AdaDelta
* RMSprop
* Adam
entre otros.

Otras ideas que se usan en DL. Son

* Stochastic Gradient Descent: Estocastico, se escoge solo un punto aleatorio de los datos para calcular el gradiente y asi el algoritmo es mas veloz pero menos preciso.

* Batch Gradient descent. Es el mismo gradiente que conocemos , sobre todos los puntos

* Minibach Gradient descent: sobre un grupo pequenho de puntos.

Para terminar una nota de **line search**:
Recuerden que el SD nace de hallar $\alpha$ que minimice
$\varphi(\alpha)= J[ \theta_0 - \alpha \nabla J(\theta_0)]$
Como $J$ era analitica, pudimos encontrar $\alpha=\|r_i \|^2/r_i^T A r_i $, pero si $J$ es altamente no lineal, puede que no halla forma analitica de encontrar $\alpha$. Que se puede hacer?

* Tratar 3 puntos distintos en la curva $\varphi$. Por 3 puntos solo pasa una parabola y halla el minimo en esa parabola.
* backtracking? Selecciona un $\alpha$ grande y se comienza a reducir. Buscando convergencia.

* comenza bajito y luego incrementa, buscando eficiencia.
