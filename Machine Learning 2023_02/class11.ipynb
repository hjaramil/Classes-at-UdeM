# Pseudoinversa (Continuacion)
Sobre la pseudoinversa. Piense en el sistema $Ax=b$,
si $A$ tiene inverso $A^{-1}=A^+$. $x = A^{-1} b$.
Si $A$ ni siquiera es cuadrada es $m  \times n$,
entonces queremos minimizar el problema
$\| Ax - b \|^2$. Esto da las ecuaciones normales $A^T A x = A^T b$, si $A^T A$ es invertible entonces

$$ x = (A^T A)^{-1} A^T $$
y $(A^T A)^{-1} A^T$ es la pseudoinversa.
Pero que pasa si $A^T A$ no es invertible? Esto ocurre
cuando las columnas de $A$ no son linealmente independientes.
Entonces aca es donde la solucion no es unica y toca hallar la normal al espacio nulo. Esta es la mas pequena de todas las $x$ que resuelven el problema


Si $b  \ne \mathcal{R}(A)$ es entonce $Ax =b$ no tiene solucion.
Podemos escribir $b = b_r + b_0$, $b_r \in \mathbb{R}(A)$.
donde

$$ \| Ax - b \|^2 = \| Ax - b_r \|^2 + \| b_o \|^2 .$$
Queremos minimizar $\min_x \| Ax - b \|^2$ la respuesta es

$$ \min_x \| Ax - b \|^2 = \| b_o \|^2.$$
El vector $b_o$ es ortogonal a $\mathcal{R}(A)$.  Es decir

$$ \langle b_0, Ax \rangle = \langle A^* b_0, x \rangle = 0  .$$
Es decir $A^* b_0 =0$, $\forall x$. En otras palabras
**$b_0$ esta en el espacio nulo de $A^*$**.
Hagamos el siguiente analisis:


\begin{eqnarray}
Ax = b_r &⇔&  \\
&\Leftrightarrow & Ax = b-b_0  \\
&⇔&  Ax -b \in \mathcal{N}(A^*) \\
&⇔& A^*(Ax -b) = 0 \\
&⇔& A^* A x = A^* b
\end{eqnarray}
y volvemos a las ecuaciones normales.  Este es el problema
de minimos cuadrados. Si $A^*A$ es invertible

$$ x = (A^* A)^{-1} A^* b .$$
Esta la solucion en minimos cuadrados. Pero que pasa si $A^* A$
no es invertible?  Hay espacio nulo. Bien sabemos del algebra lineal (de ecuaciones diferenciales) que si $x_p$ es una solucion particular de $Ax =b$, entonces $x = x_p + x_o$, donde $x_o \in \mathcal{N}(A)$ es tambien solucion, pues

$$ Ax = A( x_p + x_0) = A x_p + A x_o =Ax_p = b.$$
Cual es la mas peuquena de este numero infinito de soluciones?
Las mas pequena es $x$ tal que $\langle x, w \rangle = 0$, $w \in \mathcal{N}(A)$, o en otras palabras $x \perp \mathcal{N}(A)$.  

Analiticamente esto se prueba asi.
$$\| x \|^2 = \| x_N \|^2 + \| x_o \|^2 .$$
El mas pequeno es $x_o $ que es perpendicular al espacio nulo.





Definicion **pseudoinversa** : La **least-squares pseudo-inverse**
de la matriz $A$ es la matriz $A^+$, para la cual $x= A^+ b$ satisface dos hipotesis:

1. $$A^* A x = A^* b  \quad, \quad \text{ecuaciones normales, LS}$$
2. $$ \langle x, w \rangle = 0  \quad \text{para todo $w$ tal que $Aw = 0$}.$$
En otras palabras, la condicion 2 dice que $x \perp \mathcal{N}(A)$.
Nota: En la literarua es mas comun encontrar el concepto de
[More Penrose Inverse](https://en.wikipedia.org/wiki/Moore%E2%80%93Penrose_inverse)
Sin embargo este metodo, que produce la misma matriz, es muy diferente al que usamos aca.

El metodo que usamos aca es de James P. Keener (paso la bibliografia en el aula virtual ya:)

 James P. Keener. Principles of Applied Mathematics: Transformation and approximation.
Addison-Wesley Longman Publishing Co., Inc., Boston, MA, USA, 1988.

Ejemplo: Considere la matriz

\begin{eqnarray}
A = \begin{pmatrix} 1 & 1 \\ 1/2 & 1/2 \end{pmatrix}
\end{eqnarray}
y el vector $b = (1, 3)^T$. La matriz $A^+$ no depende de $b$.
Se usa $b$ como ayuda para encontrala.
**A mano**

Se hacen dos cosas:

1. Hallar $\mathcal{R}(A)$, proyectar el vector $b$ en el rango de $A$.
2. Hallar la ecuacion ortogonal al espacio nulo.

1. El rango de $A$ esta generado por el vector $(1, 1/2)^T$
o mejor $(2, 1)^T$ pero hay que normalizarlo

\begin{eqnarray}
u = \frac{1}{\sqrt{5}}  \binom{2}{1}.
\end{eqnarray}
Recuerden la proyeccion:


$$\langle b, u \rangle u = \frac{1}{\sqrt{5}} \left  \langle \binom{b_1}{b_2} , \binom{2}{1}  \right \rangle  \frac{1}{\sqrt{5}} \binom{2}{1} = \frac15 (2 b_1 + b_2) \binom{2}{1} =b_r $$
y $Ax = b_r$ tiene solucion.
Escribamos el sistema. Por simplificacion $\alpha = (1/5)(2 b_1 + b_2)$, entonces tenemos el sistema

\begin{eqnarray}
\begin{pmatrix}
1 & 1 \\ 1/2 & 1/2
\end{pmatrix}
\binom{x_1}{x_2}
= \alpha \binom{2}{1}.
\end{eqnarray}

Las dos ecuaciones son la misma (multiplique la segunda por 2 y obtiene la primera. Tenemos (la primera)


$$ x_1 + x_2 = 2 \alpha.$$
Esta ecuacion tiene un numero infinito de soluciones. Necesitamos hallar el espacio ortognal al espacio nulo.

El espacio nulo es $x_1 + x_2 = 0$, esto da $x_2 = - x_1$
(recuerde que la pendiente de la ortogonal es la inversa negativa, o sea $m_o=1$, entonces $x_2 = x_1$
Esta ecuacion es $x_2 - x_1 = 0$
Tenemos dos ecuaciones:

$$ x_1 + x_2 = 2 \alpha. \tag{1} $$
$$x_2 - x_1 = 0 \tag{2} $$
Resolvamos el sistema: Sume (1) + (2)
$2 x_2 = 2 \alpha$, $x_2=\alpha$, reemplazando en la segunda
entramos $x_1=x_2=\alpha$, es decir

Tenemos entonces:

\begin{eqnarray}
x_1 &=& \frac15 ( 2 b_1 + b_2) \\
x_2 &=& \frac15 ( 2 b_1 + b_2) \\
\end{eqnarray}
Se lleva a forma matricial
\begin{eqnarray}
\binom{x_1}{x_2} = \begin{pmatrix}
\frac25 & \frac15 \\
\frac25 & \frac15 \\
\end{pmatrix}
\binom{b_1}{b_2}
\end{eqnarray}













Es decir

\begin{eqnarray}
A^+ = \begin{pmatrix}
2/5 & 1/5 \\ 2/5 & 1/5
\end{pmatrix}.
\end{eqnarray}

## El uso de SVD para hallar la pseudoinversa.
Teorema (ejercicio):
Propiedades:
1. Si $A$ es $m \times n$, entonces $A^+$ es $n \times m$. Se prueba con analisis dimensional.
2. Si $A$ es invertible entonces $A^+=A^{-1}$.
3. Si $A$ es diagonal $\Sigma$, entonces $A^+$ tambien es diagonal con  las entradas $\sigma_i^{-1}$ cuando $\sigma_i \ne 0$, y 0 cuando $\sigma_i = 0$.
4. La pseudoinversa de $(AB)$ es $B^+ A^+$, o sea

$$ (AB)^+ = B^+ A^+ .$$

5. Si $A= U \Sigma V^T$ por SVD. Entonces

$A^+ = V \Sigma^+ U^*$
