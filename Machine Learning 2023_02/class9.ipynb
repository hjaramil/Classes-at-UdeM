# Minimos cuadrados con regularizacion
Asumamos $x \in \mathbb{R}^{m \times n}$, $b \in \mathbb{R}^m$, $x \in \mathbb{R}^n$. Podemos escribir $A^T$ en vez de $A^*$ pero por generalidad dejamos la notacion $A^*$.

Planteamos el siguiente problema: Resolver $Ax = b$.
Si $A$ es invertible (no singular) entonces
$$x = A^{-1} b. $$
Sin embargo esto rara vez ocurre. Basta que $m \ne n$ y no puede ocurrir. Hay dos casos importantes. Que no halla solucion, o que halla un numero infinito de soluciones.

1. Que la matriz sea "overdetermined" sobre-determinada, es decir que el sistema $Ax = b$ tiene mas ecuaciones independientes que variables. En este caso no existe solucion

\begin{eqnarray}
3 x + y &=& 5 \\
3 x - y &=& 6 \\
2 x + y &=& 7
\end{eqnarray}
Aca

\begin{eqnarray}
A = \begin{pmatrix} 3 & 1 \\ 3 & -1 \\ 2 &1 \end{pmatrix}
\end{eqnarray}
En ese caso el espacio nulo es $\mathcal{N}(A) = \{ 0 \}$
No hay solucion pro que el vector $(5,6,7)^T$ no esta en
el $\mathcal{R}(A)$. Dejamos esto como tarea.

2. El segundo es que $A$ sea subdeterminada. El sitema $Ax=b$
tiene mas variables que ecuaciones independientes

Ej: $2 x + 3y = 5$
Aca $A=(2, 3)$. En este caso el espacio nulo de $A$ no es el trivial $\mathcal{N}(A)$. Seria la solucion a $2x + 3y=0$,
caulquier vector $s(3,-2)$ es solucion.  Veamos

$$ 2 (3s) - 3(2s)= 6s - 6s = 0 .$$
$s \in \mathbb{R}$.
El siguiente ejemplo no es sobrederminado, por que las
ecuaciones no son linealmente independientes
\begin{eqnarray}
3 x + y &=& 5 \\
3 x - y &=& 6 \\
6 x + 2y &=&10
\end{eqnarray}
La tercera es 2 veces la primera. El sistema no es ni sobredeterminado ni subdeterminado. Es determinado.

Recuerden que el problema que queremos resolver es $Ax =b$.
Como no simpre tiene solucion podemos pensar en encontrar el minimos de $\frac12 \| Ax - b \|^2$. Este es el momento para describir lo que es una una funcion convexa.

Definicion (**convexa**) una funcion es convexa si al unir dos puntos sobre la grafica, el segumento que los une esta por encima de ella. **concava** si la grafica esta por encima del segmento.
Pero podemos multiplicar una concava por (-1) y se vuelve convexa.
En los reales (funciones reales) una funcion es convexa si la segunda derivada en su dominio es positiva.  Puede ser **estrictamente** convexa (o concava) si la derivada es 0 o positiva (0 o negativa).



Es posible, y lo veremos con ejemplos que halla un numero infinito
de puntos tales que $\| Ax - b \|^2=0$. Tampoco es una solucion muy buena. Esto pasa cuando el espacio nulo no es el trivial $\{0\}$,
es decir $\mathcal{N}(A) \ne \{ 0 \}$, a su vez esto pasa cuando las columnas de $A$ no son linealmente independientes. (esto viene del algebra lineal).
Si retomamos las ecuaciones normales de la clase anterior
tenemos que la minizacion de $\| Ax - b \|^2$ esta data por la solucuion de

$$A^* Ax = A^* b .$$
De forma que si $A^* A$ tiene inversa, mulplicando a ambos lados por la inversa tenemos

$$ x = (A^* A)^{-1} A^* b.$$
Esta es la solucion de minimos cuadrados.

El problema que se presenta en este momento es que pasa si
$A^*A$ no tiene inversa?
Cuando $A^*A$ no tiene inversa? (cuadrada).




Entonces en este caso queremos **regularizar** el problema. Para esto Tikhonov propuso minimizar

$$ f(x) = \| A x - b \|^2 + \| \Gamma x \|^2  \tag{1}.$$
Se busca minimizar la funcion $f$. A este problema se le conoce como **regularizacion de Tikhonov**. Ojo, donde $\Gamma$ es una matriz positiva definida.
Lo que hace la matriz $\Gamma$ que es positiva definidas es "empujando" el espectro para que no hallan autovalores $0$ de $A^* A$.

Tarea: mostrar que la solucion de $\nabla f(x) = 0$, nos produce

$$x = (A^* A + \Gamma^* \Gamma)^{-1} A^* b \tag{2} .$$
Toca explicar que la matrix  $A^* A + \Gamma^* \Gamma$
tiene inversa por que $\Gamma$ es positiva definida.
La ecuacion (2) es la solucion al problema $Ax=b$ con **minimos cuadrados regularizados**.

En la practica se usa $\Gamma= \sqrt{\lambda} I$, donde $\lambda >0$ pero $\lambda \ll 1$. Este caso la Ecuacion (2) se vuelve

$$ x = (A^* A + \lambda I)^{-1} A^* b \tag{3}.$$
A este tipo de inversion en estadistica se le llama **ridge regression**.

Vamos a llegar a (3) de una forma intuitiva.

Arrancamos de
$$ Ax = b .$$
Como $A$ no tiene inversa, entonces multiplicamos a ambos lados por $A^*$ y obtenemos

$$ A^*Ax = A^* b .$$
Nos preguntamos si $A^*A$ tiene inversa. Si si, multiplicamos a ambos lados por $(A^* A)^{-1}$ y obtenemos

$$ x = (A^* A)^{-1} A^* b .$$
Sin embargo que pasa si $A^*A$ no tiene inversa? esto quiere decir que algunos autovalores de $A^* A$, son 0. Para "correr" (empujar) es espectro al lado positivo escogemos $\lambda> 0$,
y cambiamos esta solucion por


$$x = (A^* A + \lambda I)^{-1} A^* b .$$

Estamos hablando de tres cosas:
1. Problema $Ax = b$.
2. Funcion objetivo
$$ J(x) = \frac12 \| Ax - b \|^2 + \lambda \| x \|^2 \tag{4}.$$
Obverse que esta funcion objetivo es la misma (1) con
$\Gamma = \sqrt{\lambda} I$.
3. La solucion es
$$x = (A^* A + \lambda I)^{-1} A^* b .$$

Ahora bien si $\lambda \gg 1$, el inverso es casi 0 y la solucion se acerca a 0. Debe haber un compromiso. Por ejemplo $\lambda = 10^{-8}$.


Veamos otra forma de ver este problema. La minimizacion de (4) se puede ver como un problema de **multiplicadores de Lagrange**.

En este caso escribimos el problema como

\begin{eqnarray}
\min_x \frac12 \| Ax - b \|^2 \\
\text{sujeto a la restriccion } \| x \|^2 \le c
\end{eqnarray}
donde $c$ es un numero positivo.

Ejemplos del uso de la diferenciacion usando notacion indicial de funciones cuadraticas.

Asuma $A$ es simetrica y real
$$ f(x) = \frac12 x^T A x - b^T x + c.$$
Hallar el gradiente usando indices

$$ f(x) = \frac12 \sum_{i=1}^n \sum_{j=1}^n a_{ij} x_i x_j - \sum_{j=1}^n b_j x_j + c.$$

Tomando la derivada parcial con respecto a $x_k$.
recuerde $\partial x_i/\partial x_k= \delta_{ik}.$

\begin{eqnarray}
\frac{\partial f}{\partial x_k} &=&
= \frac12 \sum_{i=1}^n \sum_{j=1}^n a_{ij} \delta_{ik} x_j  +
 \frac12 \sum_{i=1}^n \sum_{j=1}^n a_{ij} x_i \delta_{jk}
- \sum_{j=1}^n b_j \delta_{jk} \\
&=& \frac12 \sum_{j=1}^n a_{kj} x_j + \frac12 \sum_{i=1}^n a_{ik} x_i - b_k \\
&=& \sum_{j=1}^n a_{kj} x_j - b_k \\
&=& (Ax)_k - (b)_k
\end{eqnarray}
Entonces encontramos que

$$ \nabla_x f(x) = Ax - b.$$
