**Teorema Espectral 1**: Si $A$ es autoadjunta, los autovectores correspondientes a autovalores distintos son ortogonales y los autovalores son todos reales.

Asumamos que $\lambda_j$ es un autovalor correspondiente a un autovector $u_j$. Sin perdida de generalidad podemos asumir que $u_j$ es unitario. (sino, se puede normalizar y el proceso que sigue aplica igual).
Trabajamos con la forma cuadratica $\langle A u_j, u_j \rangle$.
Por que es cuadratica?

$$\langle A u_j , u_j \rangle = u_j^* A^* u_j = u_j^* A u_j $$
$$\langle A u_j , u_j \rangle = u_j^* A^* u_j = u_j^* A u_j = \langle u, A u_j \rangle$$


En general:
$$\langle A u, v \rangle = u^* A^* v = u^* (A^* v) = \langle u, A^* v \rangle   \tag{1} .$$
$\| u_j \|=1$,

Probemos primero que $\lambda_j$ es real. Recuerde que si $\alpha = \overline{\alpha}$, entonces $\alpha$ es real.

$$\langle A u_j, u_j \rangle = \langle \lambda_j u_j, u_j \rangle = \overline{\lambda_j} \langle u_j, u_j \rangle = \overline{\lambda_j}   .$$
Ahora, dado que $\langle A u_j, u_j \rangle = \langle u_j, A^* u_j \rangle .$ ver Ecuacion (1). entonces

$$ \overline{\lambda_j} = \langle A u_j, u_j \rangle =\langle u_j, A^* u_j \rangle = \langle u_j , A u_j \rangle = \langle u_j, \lambda_j u_j \rangle = \lambda_j \langle u_j, u_j \rangle = \lambda_j .$$

Tenemos entonces de la ultima ecuacion $\overline{\lambda_j} = \lambda_j$.
O sea que $\lambda_j$ es real.


Toca probar todavia que los autovectores son ortogonales.
Aca tambien usamos formas cuadraticas.
Sean $u,v$ dos autovectores correspondientes a dos autovalores
$\lambda_1, \lambda_2$ distintos.

$$ \langle Au , v \rangle = \langle \lambda_1 u, v \rangle = \lambda_1 \langle u, v \rangle  . \tag{2} $$
Ahora como $A=A^*$, entonces

$$ \langle Au , v \rangle =  \langle u, A v \rangle = \langle u, \lambda_2 v \rangle = \lambda_2 \langle u, v \rangle . \tag{3} $$

De las Ecuaciones (2) y (3)
$$ \lambda_1 \langle u , v \rangle = \lambda_2 \langle u , v \rangle .$$
Pasando a restar

$$(\lambda_1 - \lambda_2) \langle u , v \rangle = 0 .$$
pero como $\lambda_1 - \lambda_2 \ne 0$ y podemos dividir por esto y obtener

$$ \langle u, v \rangle = 0$$. O sea que $u$ es ortogonal a $v$.
o $ u \perp v$

Con este hecho podemos simplificar muchos problemas. Por ejemplo veamos la siguiente factorizacion (de matrices), o ortogonalizacion o.

Asumamos que todos los autovalores en el espacio $\mathbb{C}^n$
son distintos. Entonces podemos decir que los autovectores
$\{ x_1, x_2, \cdots, x_n \}$ son ortogonales(normalizados entre si.
Sabemos $A x_i = \lambda_i x_i$, ahora tenemos
$x_i^* x_j = \delta_{ij} $ (por la ortogonalidad) Es decir

$$ x_i^* A x_j = x_i^* \lambda_j x_j= \lambda_j \delta_{ij} . \tag{4}$$


La matriz compuesta por las columnas $x_j$, la llamamos $Q$,
Entonces en forma matricial la Ecuacion (4) se escribe como
$$ Q^* A Q = \Lambda . \tag{6}$$
donde $\Lambda$ es una matriz diagonal con los autovalres de $A$.







Si en la euacion (6) multiplicamos a ambos lados por $Q$ a la izuierda y $Q^*$ a la derecha (recuerde que $Q^* Q = Q Q^* = I$

$$Q Q^* A Q Q^* = Q \Lambda Q^* .$$
$$A = Q \Lambda Q^* . \tag{7}$$
Esta ecuacion muestra que la matriz autoadjunta $A$ se puede factorizar como el producto de una matiz ortogonal $Q$, una diagonal (con autovalores) y la adjunta de la ortogonal orignal.
Las columnas de $Q$ son autovectores normalizados de $A$.

Elevar $A^n$ es un problema elemental usando la identidad (7).

Pero antes debemos introducir el concepto de **espacio invariante** y una propiedades de este.

Un espacio lineal es el generado de alguna base (Algebra lineal 1)
Un recta, un plano, un hiperplano, todos pasan por el origen.

Definicion **espacio invariante**: Sea $W$ un espacio lineal de
dimension $n$. Decimos que $W$ es invariante para la matriz cuadradas $A \in \mathbb{C}^{n \times n}$, si $AW \subset W$.


Ejemplos de espacion invariantes:
* Los espacios triviales $\{0 \}$, $\mathcal{U} = \mathbb{C}^n$
son invariantes para cualquier matriz $A \in \mathbb{C}^{n \times n}$. Pues $A 0 = 0$, $0 \in \{ 0 \}$, y $A u = v$, pero $u \in \mathbb{C}^n$  y $v \in \mathbb{C}^n$.

* Sin embargo estos no son muy interesantes. el espacio nulo  de $A$, $\mathcal{N}(A)$ es invariante . Pues para cuaquier vector $v \in \mathcal{N}(A)$, $A v = 0$, $0 \in \mathcal{N}(A)$.

* Otro ejemplo. Cualquier autoespacio es un espacio invariante. Que es el autoespacio? Asumamos solo  un autovector $v$. El autoespacio correpondiente a este autovector es el conjunto
$\lambda v$ para $\lambda \in \mathbb{C}$.  Para un autovector $v$ el autoespacio es la recta que lo contiene (pasa por el origen). Veamos por que este autoespacio es invariante

Prueba formal:
Sea $W = s v$ ($s \in \mathbb{C}$) el autoespacio (recta) generado por $v$.
Probamos que $A W \subset W$ (esto es la definicion de invariante).

Prueba: Se toma un $u \in AW$ y mostramos que $u \in W$.
Si $u \in AW$ es por que existe un $x \in W$ , tal que
$u = Ax$, pero si $x \in W$, $x = \alpha v$, con $v$ autovector en cuestion. O sea que

$$u = A (\alpha x) = \alpha A x = \alpha (\lambda v)$ = s v $$
donde $s=\alpha \lambda$, entonces $sv \in W$
O sea que $A W \subset W$, y $W$ es invariante.

Similarmente el generado de un conjunto de autovectores (auto-espacio) es tambien invariante. Este generado lo defino como

$$ W = \langle x_1, x_2, \cdots,  x_k \rangle = \left \{ \sum_{i=1}^k \alpha_i x_i   | \alpha_i \in \mathbb{C} \right \}  $$
Dejo como prueba al estudiante probar que $AW \subset W$.

El siguiente teorema (rarito) lo necesitamos para probar el teorema de descomposicion espectral 2.

Lemma : Para cada espacio invariante $W \ne \{ 0 \}$ de una matriz $A$ existe por lo menos un autopar $\lambda, x$.

Prueba: Asumamos que la dimension del espacio $W$ es $k$, y
$\{ x_1, x_2, \cdots, x_k \}$ es una base para $W$. Cualquier matriz $A$ tiene un auto par. Lo que necesitamos probar que en $W$ existe un autvector de $A$. Recuerde que $AW \subset W$
por la definicion de invariante.

Tenemos para $x \in W$, $Ax \in W$, $x = \sum_{i=1}^k \alpha_i x_i$, con $A x_i = \sum_{j=1}^k \beta_{ji} x_j$
Necesitamos probar que la ecuacion $Ax - \lambda x = 0$ tiene solucion.

Es decir, necesitamos encontrar una solucion para

$$ Ax = \sum_{i=1}^k \alpha_i A x_i = \lambda \sum_{i=1}^k \alpha_i x_i.$$

Usaremos una identidad
$$ \lambda x_i = \lambda \sum_{j=1}^k \delta_{ij} x_j .$$

Pero esto es: Pasando a restar a la izquierda


\begin{eqnarray}
\sum_{i=1}^k \alpha_i \left ( \sum_{j=1}^k \beta_{ji} x_j - \lambda x_i     \right ) &=& 0 \\
\sum_{i=1}^k \alpha_i \left ( \sum_{j=1}^k \beta_{ji} x_j - \lambda \delta_{ij} x_j     \right ) &=& 0 \\
\sum_{i=1}^k \alpha_i \left ( \sum_{j=1}^k \beta_{ji}  - \lambda \delta_{ij}      \right ) x_j &=& 0 \\
\sum_{j=1}^k  \left ( \sum_{i=1}^k \alpha_i (\beta_{ji}  - \lambda \delta_{ij})      \right ) x_j &=& 0 \\
\end{eqnarray}
Como todos los $x_j$ son linealmente independientes los coeficientes de los $x_j$ son todos 0.

Es decir
$$ \sum_{i=1}^k \alpha_i ( \beta_{ji} - \lambda \delta_{ij}) = 0 .$$

En forma matricial estos $ (B - \lambda I) \alpha = 0$ donde $B=(\beta_{ji}$). Esto lo que muestra es que $\lambda$ si existe y es un autovalor de la matriz $B$.
