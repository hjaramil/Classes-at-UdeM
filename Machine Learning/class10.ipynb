# Counting: Conteo.
Esto es parte de combinatoria.

Dividmos el programa en dos partes: Sets y Multisets (conjuntos y multiconjuntos).

La regla de la multiplicacion.  Se tienen $b_1$ en una caja, $b_2$ en otra y asi susesivamente hasta $b_k$ en la caja $k$.
Selection_001.png


Si por cada $b_1$ tengo $b_2$ elementos, y por cada $b_2$ tengo $b_3$ elementos y asi suscesivamente hasta que por cada $b_{k-1}$ tengo $b_k$ elementos el total es

$$ b_1 b_2 \cdots b_k = \prod_{i=1}^k b_i $$ 

Ejemplo:

* Coordenadas cartesianas en $\mathbb{R}^k$ 
$(x_1, x_2, \cdots, x_k)$. Si cada coordenada $x_i$ tiene $n_i$ posibilidades tenemos un total de
$$ \prod_{i=1}^k n_i $$

* Numeros con digitos. Pregunta: cuantos numeros se pueden escribir con 3 cifras si cada cifra puede recorrer 10 digitos $0,1,2, \cdots, 9$. 
Si $b_i=c$, 

$$ b_1 b_2 \cdots b_k = \prod_{i=1}^k b_i =c^k$$ 
Respueta $10^3=1000$, $0,1,2, \cdots, 999$.

* Contar placas de carro en USA: 2 letras y dos numeros. Por emplo $AF-231$. 
El total es $26 \times 26 \times 10 \times 10 \times 10 = 676000$.

## Conjuntos: 
Coleccion finita no es vacio, pero tampoco tiene un numero infinto de elementos. Normalmente en los conjuntos no importa el orden. Tampoco se cuentan los elementos repeditos. 

$$ \{ 1, 3 \} = \{ 3, 1 \}$$
Pero nosotros, para poder entender el conteo , vamos a consideraer dos casos.

El contexto es: Se tienen $n$ objetos en un bolsa y se van 
a extaer $k$ objetos $k < n$. 

* El orden importa (permutaciones) 
    * Con repeticion: $\prod_{i=1}^k n_i$ y si cada caja tiene los mismos elementos $n^k$.

    * Sin repeticion: (sin reemplazo)
    $$n \times (n-1) \times \cdots \times (n-k+1) = \frac{n!}{(n-k)!} = \frac{n(n-1)(n-2) \cdots (n-k+1) \cancel{(n-k)!}}{\cancel{(n-k)!}}$$

$$ conteo = \frac{n!}{(n-k)!} = n^{\underline{k}} $$
A esto se le llama el "falling factorial" (factorial cayendo)

Ejemplo: Considere el alfabeto ingles de 26 letras. De cuantas formas se pueden sacar 3 letras sin reemplazo e importando el orden

$$26^{\underline{3}}=26 \times 25 \times 24 = 15600 $$

Ahora si $k=n$, $conteo = n \times n-1 \times n-2 \times 1 = n!$

* El orden no importa (combinaciones). Como el orden no importa
$ab=ba$, $abc=acb=bac=bca=cab=cba$ ($6 = 3!$).
Por cada grupo de $k$ elementos que extraje de la bosa, yo conte el grupo $k!$ veces. 

$$conteo =   \frac{n!}{k! (n-k)!} = \binom{n}{k} =\frac{n^{\underline{k}}}{k!}$$


Cuando se expande $$(a+b)^n = \sum_{i=0}^n \binom{n}{i} a^i b^{n-i} $$ 

la expresion $\binom{n}{k}$ indica el numero de veces que podemos tomar $k$ elementos de $n$ sin repeticion y sin importar el orden. El numero de subconjuntos de un conjunto con $n$ elementos y cada subconjunto con $k$ elementos.

Una forma , mas formal, de definir el multiset es como una tupla $(A, m)$ donde $A$ es un conjunto (sin repeticiones) y $m$ es la multiplicidad:

\begin{eqnarray}
m : A &\to& \mathbb{N} \\
 x  &\mapsto& m(x) = \text{ cuantas veces esta el caracter $x$ en el multiset}
\end{eqnarray}

* Importa el orden: Pensemos en la palabra "bookkeeper".
La pregunta es: Cuantas palabras distintas se pueden escribir con las letras de la palabra "bookkeeper".

El total de letras son $n=10$. Si todas las letras fueran
distintas $n!$ . Si embargo, por ejemplo mire la letra "o".  Como la "o" esta repedita decir $o_1kero_2$ es lo mismo
que $o_2kero_1$ la estoy contando dos veces $2!$.
Mas critico es el caso de la $e$, 
$e_2okr$, $e_1okr$, $e_3okr$, hay $3!$ formas de contar las 3 $e's$ del conjunto. O sea 6 veces mas palabras de las que son y asi con las demas

$$conteo = \frac{10!}{3! 2! 2! 1! 1! 1!} =151,200$$

La formula es

$$ conteo = \frac{n!}{m_1! m_2! \cdots m_k!} = \frac{m_1 + m_2 + \cdots m_k!}{m_1! m_2! \cdots m_k!}$$

A este numero se le llama el mulinomial. Por que?

$$(x_1 + x_2 + \cdots + x_k)^n = \sum_{i_1 + i_2 + \cdots + i_k=n}  c_{i_1 i_2 \cdots i_k} x_1^{i_1} x_2^{i_2} \cdots x_{k}^{i_k}  $$

$$c_{i_1 i_2 \cdots i_k} = \binom{n}{i_1, i_2, \cdots, i_k} = \frac{n!}{i_1 ! i_2 ! \cdots i_k! } $$

Reduzcamos esto al caso de solamente 2 elementos

$n=m_1 + m_2$. 

\begin{eqnarray}
\frac{m_1 + m_2}{m_1! m_2!} = \frac{n!}{m_1 m_2} = \frac{n!}{(n-m_1)! m_2} = \binom{n}{m_1} 
\end{eqnarray}

Un problema: Cuantos subconjuntos tiene un conjunto de $n$ elementos.

$$ (a+b)^n = \sum_{i=0}^n \binom{n}{i} a^i b^{n-i} $$
Si $a=b=1$.
$$ 2^n = \sum_{i=0}^n \binom{n}{i} $$
La respuesta es $2^n$.

Recuerde el modelo de las cajas. Dado un subconjunto de
$k$ elementos  un conjunto de $n$ elementos. Tomamos 
$n$ cajas. A cada caja se le asigna un elemento del conjunto. Y preguntamos en esta casa el elemento del conjunto esta si (1) o no (0) en el subconjunto. 
O sea que cada caja tiene solo 2 valores y como son $n$ cajas  por el problema del conteo multiplicativo, al inicio de la clase, la solucion es $2^n$.

* El orden importa. Usamos un metodo que se llama stars and bars.

[stars and bars](https://en.wikipedia.org/wiki/Stars_and_bars_(combinatorics))



El problema que resolvemos ahora es el siguiente:
(3 formas distintas del mismo problema)

1) El numero de soluciones de la ecuacion
    $$x_1 + x_2 + \cdots + x_k=n$$.

2) El numero de combinaciones de $k$ objetos extraidos de una bolsa de $n$ objetos

3) El numer de veces que $k$ objetos identicos se puden meter a $k$ cajas.

Pensemos en el numeral 1)

Tomemos un ejemplo . De cuantas formas podemos sumar los numeros $0,1,2$ para obtener $x_1 + x_2 + x_3 = 2$.

\end{eqnarray}

Tenemos 4 simbolos (2+2).Y dos grupos de 2.

$$conteo = \binom{4}{2} = \frac{4!}{2! 2!} = 6 $$

Generalizamos

$x_1 + x_2 + \cdots x_k = n$.

ejemplo:
$$ || + ||| + \cdots + || + | $$
Tenemos $n$ barras, pero tenemos $k-1$ signos "+"

La formula es de cuantas formas podemos agrupar $n$ de $n+k-1$. 

La formula es:

$$ \frac{(n+k-1)!}{n!(k-1)!} = \binom{n+k-1}{n} = \binom{n+k-1}{k-1}$$

$$ \left ( \binom{n}{m} \right)=\binom{n+k-1}{n} = \frac{(n+k-1)!}{k! (n-1)!}  = \frac{n(n+1)(n+2) \cdots (n+k-1)}{k!} = \frac{n^{\overline{k}}}{k!}$$




# ANN: Artifical Neural Networks (redes neuronales artificiales)

Las redes neuronales tienen un soporte matematico fuerte.
Y es cualquier funcion (decente) de la forma

$$f: \mathbb{R}^n \to \mathbb{R}^m $$
se puede aproximar tanto como quiera con una red neuronal
de profundidad $d=\max \{n+1, m\}$ (la profundidad es el numero de capas).

 A este teorema se le llama
 [universal approximator](https://en.wikipedia.org/wiki/Universal_approximation_theorem)

 ## Introduccion. 
 Supongamos que teneos una imagen de $100 \times 100$ pixeles.
 Quisieramos ver la posibilidad de interpretar la imagen usando regresion logistica. Clasficar: Es perro, es gato , es cerdo.
 Podriamos incluso tener la imagen a color (RGB), hay otros formatos (HSV: Hue Saturation Value), HSL=Hue saturation lightness, CMYK: Cyan, Magenta, Yellow, K=black. Pero por simplicidad vamos a usar grises. Un solo canal.
 Cada pixel es una variable. Tenemos entondes $100 \times 100 = 10,000$ variables. Piense en el poliomio no lineal mas simple (cuadratico) en $n$ variables $x_1, x_2, x_3, \cdots, x_n$. Tenemos muchos "features" (propiedades) escribimos el polinomio como sigue


 $$\theta_0 + \theta_1 x_1 + \cdots + \theta_n x_n + \theta_{n+1} x_1 x_2 + \cdots + \theta_m x_{n-1} x_n + \theta_{m+1} x_1^2 + \theta_{m+2} x_2^2 + \cdots + \theta_q x_n^2$$.





Calculamos el numero de terminos que tiene este polinomio.

Los objetos son $1,x_1, x_2, \cdots x_n$. Cualquier termino 
de este polinomio es de la forma

$$1^{i_0} x_1^{i_1} x_2^{i_2} \cdots x_n^{i_n}  $$
Todos los exponentes deben sumar $2$.

Por ejemplo $\theta_0$ se obtiene con el polinomio igual 1
En este caso $i_0=2$, $i_1=i_2= \cdots i_n=0$. 

 Otro ejemplo $x_1$. $i_1=1$, $i_2=i_3 = \cdots i_n=0$,
 $i_0=1$. 

 Ultimo ejemplo $x_2^2$. $i_2=2$, $i_0=i_1=i_2=\cdots i_n=0$.

 $x_3 x_4$: $i_3=i_4=1$, $i_0=i_1=i_2=i_5=i_6 \cdots i_n=0$.

 El problema se reduce (el numero de terminos del polinomio)
 en hallar el numero de soluciones de la ecuacion

 

 $i_0 + i_1 + \cdots i_n = 2$ (Diofantina, por que las soluciones son enteras. 
 Se reduce al probla de conteo de la clase anterior.

 El problema se reduce a

 $$\binom{n+k}{k} = \binom{n+k}{n}$$

 Veamos un ejemplo simple.

 Un polinomio en 2 variables de orden 2. Cuantos terminos tiene.

 * El metodo "bruto". 
 $$\theta_0 + \theta_1 x_1 + \theta_2 x_2 + \theta_3 x_1 x_2 + \theta_4 x_1^2 + \theta_5 x_2^2  $$
 contando estos terminos da 6.

 * Metodo inteligente: $n=k=2$. 

 $$ \binom{n+k}{k} = \binom{4}{2} = \frac{4!}{2! 2!} =6 $$

 Pensemos en la imagen con 10,000 pixeles. Estas son variables.

 El polinomio no lineal mas pequeno tiene orden 2. 
 $n=10,000, k=2$. Encontramos

 $$ \binom{n+k}{k} = \binom{10,002}{10,000} =\frac{\cancel{10000!} (100001)(100002)}{\cancel{10000!} 2!}=50,015,001$$

 Si el polinomio fuera de grado 3

 $$conteo = \binom{10003}{3} = \frac{10003!}{10000! 3!} = \frac{(10003)(10002)(10001)}{6}= 166,766,685,001.0$$ es del orden de $10^2$. El numero de terminos del polinomio. Pero:

 * La matriz es de tamano $m \times n$, $m$ es el numero de terminos del polinomio . Usualmente $m > n$ entonces la matriz
 seria mas grande que $m \times n > n \times n > 10^{24}$.
