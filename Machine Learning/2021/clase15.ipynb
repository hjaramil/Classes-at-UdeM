# Multiplicadores de Lagrange


\begin{eqnarray}
\text{minimizar   } J(w) = \frac{ \| w \|^2 }{2} \\  
\text{sujeto a   } y_i (w^T x_i - b) \ge 1 \quad , \quad i=1,2, \cdots, , m
\end{eqnarray}
O equivalentemente
\begin{eqnarray}
\text{minimizar  } J(w) = \frac{\| w \|^2}{2}  \\  
\text{sujeto a  }  [ y_i( w^T x_i - b ) - 1 ] \ge 0 \quad , \quad i=1,2, \cdots, m 
\end{eqnarray}

Este problemas son de la forma:     
\begin{eqnarray}
\min f(\theta)  \quad , \quad \text{sujeto a condiciones} \\  
g_i(\theta) \ge 0 
\end{eqnarray}

En estos problemas usamos el metodo (multiplicadores de Lagrange)
$$ L(\theta, \lambda) = f(\theta) - \sum_i \lambda_i g_i(\theta) $$

En nuestro caso particular tenemos:      
$$ L(w, b, \lambda) = \frac12 \| w \|^2 - \sum_i \lambda_i  [y_i(w \cdot x_i - b)-1 ]  $$

Para minimizar esta funcion objetivo debemos encontrar los puntos estacionarios. Son aquellos donde $\nabla_w L = 0$.

Mas concretamente
$$\nabla_w L(w, b,  \lambda) =  \nabla \left [\frac12 \| w \|^2 - \sum_i \lambda_i  [y_i(w \cdot x_i - b)-1 ]
  \right ]  = 0 $$

Trabajemos por partes primero el primer termino y luego el segundo :
(para nosotros $\theta=(w, b)$.
$$ \| w \|^2 = \sum_i w_i^2$$ 

* Para el primer termino :     
 $$ \frac{\partial }{\partial w_j} \left ( \frac12 \| w \|^2  \right ) = \sum_i \delta_{ij} w_i = w_j$$

* Para el segundo termino:    
 \begin{eqnarray}
 \frac{\partial}{\partial w_j }  \left (    \sum_i \lambda_i  [y_i(w \cdot x_i - b)-1 ]
   \right ) &=&   \frac{\partial }{\partial w_j}  \left ( \sum_i \lambda_i \left [ y_i \left ( \sum_k w_k (x_i)_k  - b \right ) -1  \right ]
     \right )  \\ 
     &=& \sum_i \lambda_i y_i \sum_k \left [ \delta_{kj} (x_i)_k 
       \right ]  \\  
       &=& \sum_i  \lambda_i y_i (x_i)_j
\end{eqnarray}

Considerando ambas partes:$$\nabla_w L(w, b, \lambda) = w_j -   \sum_i  \lambda_i y_i (x_i)_j = 0$$

En forma vectorial:  

$$w = \sum_i \lambda_i y_i x_i $$

Ahora debemos ejecutar la derivada con respecto a $b$.


$$\frac{\partial }{\partial b} \left (\sum_i \lambda_i  [y_i(w \cdot x_i - b)-1 ]
  \right ) = -\sum_i  \lambda_i  y_i = 0 $$

En conclusion
$$ \sum_i \lambda_i y_i = 0 $$





\begin{eqnarray}
L(w, b, \lambda) &=& \frac12 \| w \|^2 - \sum_i \lambda_i  [y_i(w 
\cdot x_i - b)-1 ]    \\ 
&=& \frac12 \left (  \sum_i \lambda_i y_i x_i  \right ) \cdot 
\left (  \sum_i \lambda_i y_i x_i  \right ) - \sum_i \lambda_i  \left [y_i \left ( \sum_j \lambda_j y_i x_j  \cdot x_i - b  \right ) -1
  \right ]  \\   
  &=& \frac12 \sum_i \sum_j \lambda_i \lambda_j y_i y_j x_i \cdot x_j - \sum_i \sum_j \lambda_i \lambda_j y_i y_j x_i \cdot x_i - b \sum_i \lambda_i y_i  + \sum_i \lambda_i 
\end{eqnarray}

En resumen#

$$ L(\lambda) = -\frac12 \sum_i \sum_j \lambda_i \lambda_j y_i y_j x_i \cdot x_j   + \sum_i \lambda_i $$
Suejeto a:     

$$\sum_i \lambda_i y_i = 0$$

Este se llama el **dual** .  En el problema original los parametros eran $w, b$, ahora son $\lambda_i$ y queremos resolver

$$ \frac{\partial L}{\partial \lambda_i} = 0  \quad , \quad i=1,2, \cdots , m $$


# The Kernel Trick

faltando 10 para las 9.

Repaso del principio de la clase. Que es el Kernel. Tenemos puntos en una dimension $x_i$, construimos puntos en
dos dimensiones $(x_i, x_i^2)$. En mas dimensiones. $(x_i, y_i) \to \Phi(x_i, y_i) =  (x_i, y_i, x_i^2  + y_i^2) $. En general los puntos (vectores) $x_i$ los puede
mapear a una una dimension superior mediante una funcion $\Phi$. 
Si se mapean esos puntos y se quiere utilizar la tecnica derivada arriba. 


$$L(\lambda) = - \frac12 \sum_i \sum_j \lambda_i \lambda_j y_i y_j \Phi(x_i) \cdot \Phi(x_j) + \sum_i \lambda_i $$

El tipo de productos internos $\Phi(x_i) \cdot \Phi(x_j)$ se encuentran muy a menudo. Por ejemplo , las condiciones que se necesitan para que los puntos esten en el lado correcto de la frontera son#
$$\sum_j \lambda_j y_j  \Phi( x_j) \cdot \Phi(x_i )- b \ge 1 , \quad y_i = +1 $$
$$\sum_j \lambda_j y_j \Phi(x_j) \cdot \Phi(x_i) - b < - 1 , \quad y_i = -1  $$

Definimos el kernel como
$$ k(x_i, x_j ) = \Phi(x_i ) \cdot \Phi(x_j) $$
a $\Phi$ se le llama **feature map**

Vamos a ver ejemplos donde el kernel se pude simpliciar mediante lo que llaman el **kernel trick**






* Tomeoms 2 dimensiones $n=2$. Definamos el **feature map**

\begin{eqnarray}
\Phi : \mathbb{R}^2 &\to& \mathbb{R}^3 \\
(x_1, x_2) &\mapsto& \Phi(x_1, x_2) = (x_1^2 , x_2^2, \sqrt{ 2 x_1 x_2}) 
\end{eqnarray}

Definamos el kernel como  $k(x,y) = (x \cdot y)^2$. Vamos a probar que
$k(x,y) = \Phi(x) \cdot \Phi(y)$. 

Tomemos $x=(x_1, x_2)$, $y=(y_1, y_2)$ entonces
$$k(x,y) =(x \cdot y)^2 = \left ( \sum_{i=1}^2 x_i y_i \right ) = x_1^2 y_1^2 + 2 x_1 y_1 x_2 y_2 + x_2^2 y_2^2 $$

Ahora bien,
$$\Phi(x) \cdot \Phi(y) = (x_1^2, x_2^2, \sqrt{2 x_1 x_2}) \cdot (y_1^2 , y_2^2, \sqrt{2 y_1 y_2}) =  x_1^2 y_1^2 + 2 x_1 x_2 y_1 y_2 + x_2^2 y_2^2  $$ 


Contemos operaciones: $\Phi(x) \cdot \Phi(y) $. 4 cuadrados, 4 productos, 2 raices cuadradas, 3 productos y 2 sumas:  Total 7 productos, 4 cuadrados, 2 raices cuadradas y 2 sumas.

Ahora bien para $k(x,y)=(x \cdot y)^2$: 2 productos una suma y un cuadrado.

*  Para n=3 Definamos:    

\begin{eqnarray}
\Phi : \mathbb{R}^3  &\to& \mathbb{R}^6  \\
(x_1, x_2, x_3) &\mapsto& (x_1^2, x_2^2, x_3^2, \sqrt{2} x_1 x_2, \sqrt{2} x_1 x_3, \sqrt{2} x_2 x_3 ) 
\end{eqnarray}

Definimos $k(x,y) = (x \cdot y)^2$. Entonces#
$$k(x,y) = (x \cdot y)^2  = \left ( \sum_{i=1}^3 x_i y_i \right )^2 = \sum_{i=1}^3 \sum_{j=1}^3  x_i x_j y_i y_j  $$

Ahora bien:      

\begin{eqnarray}
\Phi(x) \cdot \Phi(y) &=& (x_1^2, x_2^2, x_3^2, \sqrt{2} x_1 x_2, \sqrt{2} x_1 x_3, \sqrt{2} x_2 x_3) \cdot (y_1^2, y_2^2, y_3^3, \sqrt{2} y_1 y_2, \sqrt{2} y_1 y_3, \sqrt{2} y_2 y_3) \\  
&=& \sum_{i=1}^3 \sum_{j=1}^3 x_i x_j y_i y_j
\end{eqnarray}

Contemos opearciones. $\Phi(x) \cdot \Phi(y)$. 

*  6 cuadrados
*  12 productos
* 1 raiz cuadrada 
* 6 productos
* 5 sumas

Contra
$k(x,y)=(x \cdot y)^2$

* 3 productos
* dos sumas
* un cuadrado
