El algortimo de K-means es el siguiente

1. Se decide cuantos clusters $k$ se requieren, obviamente $k<m$ donde $m$ es el numero de datos

2. De forma aleatoria se crean $k$ "centros". Cada centro se escribe $\mu_j$, $j=1,2, \cdots , k$. Cada centro esten en $\mathbb{R}^n$

3. Se asignan a cada centro los puntos mas cercanos a el. Se inicializa un vector $c$ con $m$ elementos, donde cada $c_i$ es un indice en rango $1,2, \cdots , k$ indicando el centro que le correspondte. Un ejemplo. Digamos que tenemos 8 puntos, y $c$ es el vector $c=[1,1,2,3,1,3,2,2]$, esto indica que los puntos asociados al centro $\mu_1$ son $x_1, x_2, x_5$, los puntos asignados al centro $\mu_2$ son $x_3, x_7, x_8$ y finalmente los puntos asignados al centro $\mu_3$ son $x_4 , x_6$

4. Se redefinen los nuevos centros:     
\begin{eqnarray}
  \mu_1 = \frac{x_1 + x_2 + x_5}{3}  \\ 
  \mu_2 = \frac{x_3 + x_7 + x_8}{3}  \\ 
  \mu_3 = \frac{x_4 + x_6}{2}
\end{eqnarray}

En este momento nos devolvemos al punto 3. e iteramos con el 4. (3., 4., 3., 4., 3. 4. ... )

Cuando se para el algoritmo . De alguna forma podemos poner un limite para el numero de iteraciones pero existe otro criterio.

$$ J(x,c,\mu) = \sum_{i=1}^m \| x_i - \mu_i \|^2   $$

$$ J(x,c, \mu) = \sum_{j=1}^k  \sum_{i=1}^{| c |} \| x_{c_i} - \mu_{c_i}  \|  $$

En la primera expresion $\mu_i$ es el centro que le corresponde a $x_i$, cuantos $\mu_i$ hay?   hay $k$

## Como sabemos cuantos $k$ son efectivos

Debemos buscar metodos que cumplan las siguientes condiciones#

* Que los objetos del mismo cluster esten tan juntos como sea posible
* Los objetos de clusters distintos esten lo mas alejados posibles


Los metodos para hallar $k$ son basicamente:      

* Metodos directos: El metodo de la silueta, el metodo del codo (elbow)  
* Metodos estadisticos: Por ejemplo "gap statistic method"


### Metodo del codo (el elbow)

Se piensa la funcion objetivo $J(x,c, \mu)$ como funcion solamente de $k$. Es decir $J(k)$ donde $k$ es el numero de particiones (de clusters)
El maximo numero de $k$ es $m$ pero si hay $m$ clusters $J(m) =0$.
Ya sabemos que $J(k) \to 0$ a medidas que $k \to m$.

Sin embargo el metodo del codo no es muy preciso, se requiere de algo mas facil de cuantificar.

### El metodo de la silueta

# New Section

Recordando algo que dijimos antes: Buscar que los puntos de un cluster esten tan juntos como se pueda (cohesion) y los de distintos cluster tan alejados como se pueda (separacion).

Se definen los siguientes conceptos#

$$ a(i) = \frac{1}{|C_i|-1 } \sum_{j \in c_i, i \ne j} d(i,j) \quad , \quad | c_i | > 1  $$
En palabras esta es la distancia media entre los puntos del cluster con centro $\mu_i$.
Otro concepto donde se toman las distancias medias a puntos en clusters ajenos. 
En ingles "dissimilarity" 
$$ b(i) = \min \frac{1}{| C_i |} \sum_{j \in C_i} d(i,j) $$


Ahora con los conceptos de cohesion y "dissimilarity" se puede definir la silueta (silhouette)

\begin{eqnarray}
s(i) = \left \{ 
    \begin{array}{c}
    \frac{b(i)-a(i)}{\max \{ a(i), b(i) \}} & \text{si  } |C_i |>1 \\ 
    0  & \text{   si}  |C_i | = 1
    \end{array}
    \right . 
\end{eqnarray}

La silueta $-1 \le s(i) \le 1$.

Si normalizamos los datos restandole la media tenemos
$$ \| x \|^2 = \| x - \overline{m}\|^2 + \|  \overline{m} \|^2  $$

No importa el angulo $\theta$, el lado izquierdo no cambia. 
Tenemos inicialmente la funcion objetivo:

Minimzar $$ J_0(\theta)  = \sum_{i=1}^m \| x_i - \overline{m}_i \|^2$$

Pero como la suma de $\| x - \overline{m} \|^2 + \| \overline{m}_i\|^2$ es constante entonces minimizar la suma de arriba es lo mismo que **Maximizar**

$$ J(\theta) = \sum_{i=1}^m \| \overline{m}_i \|^2$$

No estamos hablando de regresion por dos razones:

Recuerde que $\| m_i \|$ es la proyeccion de $x$ en la linea de direccion $u$. De esta forma



$\| m_i \| = \langle x_i, u \rangle $. 
Ahora bien escribamos el vector

\begin{eqnarray}
\begin{pmatrix}
\langle x_1 , u \rangle \\ 
\langle x_2, u \rangle \\  
\vdots \\   
\langle x_m, u \rangle
\end{pmatrix} = X u
\end{eqnarray}
with 
\begin{eqnarray}
X = \begin{pmatrix}
x_1 \\ 
x_2 \\ 
\vdots \\ 
x_m
\end{pmatrix}
\end{eqnarray}

Entonces
\begin{eqnarray}
\| Xu \|^2 = \sum_{i=1}^m \langle x_m, u \rangle^2 = \sum_{i=1}^m \| \overline{m}_i \|^2 
\end{eqnarray}

y 
$$J(\theta) =  \| X u \|^2 = u^T X^T X u $$


El quotient Rayleigh (el cociente de Rayleigh) de una matriz $A$ es $x^T A x$. Es una forma cuadratica. Es una hiperelipsoide ($A$ es simetrica y postiva definida).  El maximo valor del cociente de Rayleigh es el maximo autovalor $\lambda_1$ de $A$. En este caso $A=X^T X= \Sigma$. que es la matriz de covarianza de los datos.  Si $x$ no fuera unitario el cociente de Rayleigh seria $x^T A x/ \| x \|^2$. El autovalor mas grande de $X^T X$ es $\lambda_1$ y es $\max_{\theta} J(\theta) = \lambda_1 $. Correspondiente a $\lambda_1$ esta el autovector $u_1$. 

El autovector correspondiente a $\lambda_1$ es la componente principal. 

Ahora pensamoe en problema (cociente de Rayleigh)

$$\max_{\| u \|=1, u \perp u_1} u X^T X u $$
Se puede probar (es un ejercicio para el estudiante), que este maximo corresponde al segundo autovalor de $X^T X$. La direccion es aquella del segundo autovector.  Si seguimos asi (eliminando espacioes y usando el espacio ortogonal restante) encontramos una secuencia de autovalres

$$ \lambda_1, \lambda_2, \cdots , \lambda_n $$, junto con los autvectores $$ u_1, u_2, \cdots, u_n $$. A estos autovectores se les denomina componentes principales (PCA: Princiapla componente Analysis).

A este analisis en esta forma se le llama el MinMax principle (Courant).


Lo interesante de todo esto es que esto es SVD disfrazado. 
Singular Value Decomposition.









## PCA usando SVD:




Por SVD, $X=U \Lambda V^T$ donde $U \in \mathbb{R}^{m \times m}$ esta formado por columans que son autovectores de $X X^T$ donde $\Lambda \in \mathbb{R}^{m \times n}$ es una matriz diagonal con entradas $\sigma_i^2$ donde cada $\sigma_i^2$ es un autovalor de $X^T X$, y $V$ es una matriz cuyas columnas son los autvectores de $X^T X$. 

Ahora

$$X^T X = (V \Lambda^T U^T)(U \Lambda V^T) = V \Lambda^2 V^T $$ 

Donde $\Lambda^2$ es una matrz $n \times n$ (resultado de $\Lambda^T \Lambda$. Esta matriz $\Lambda^2$ tiene los autovalores de $X^T X$ .

## Reduccion de dimensionalidad.

Tenemos que los valores singulares $\sigma_1 \ge \sigma_2 \cdots \ge \sigma_n $.  Se debe fijar un valor $k$ a partir del cual $\sigma_j$, $j \ge k$ es muy pequeno comparativamente (mas adelante estudiamos el criterio para esto). Una vez fijado este $k$ **recortamos** las dimensiones desde 1 hasta $k$. Es decir, que definimos una nueva matriz **recortada**

$$W = [ V_1 | V_2 | \cdots | V_k ]  $$ $k$ preferiblemente $k < n$.










Debemos proyejctar cada $x_j$, en la direccion $v_i$ ($j=1, \cdots , m$, $i=1, \dots , n$). Esas proyecciones se escriben

$$ z_{ji} = \langle x_j, v_i \rangle  $$

A esta matriz de proyecciones la llamanos $Z$.

$$Z = X W  \quad \quad (*) $$

Ahora $Z \in \mathbb{R}^{m \times k}$, con $X \in \mathbb{R}^{m \times n}$ y $W \in \mathbb{R}^{n \times k}$.

## Como  se determina $k$.

* Promedio de error proyectado:   Podemos medir el error con respecto a la aproximacion usando la norma_2 como sigue

$$ \epsilon = \frac1m \sum_{i=1}^m \| x^{(i)} - x^{(i)}_{\text{approx}} \|^2 $$

Asumir una toleranci (por ejemplo $10^{-5}$ y parar $k$ cuando $\epsilon < \text{tol}$

* Error relativo con respecto a la variacion total. La varacion total esta dada por

$$\text{Tol}_{\text{var}} = \frac1m \sum_{i=1}^m \| x^{(i)} \|^2  $$

El error relativo esta dado por

$$ \frac{\epsilon}{\text{Tol}_{\text{var}}} =  \frac{\frac1m \sum_{i=1}^m \| x^{(i)} - x^{(i)}_{\text{approx}} \|^2}{\frac1m \sum_{i=1}^m \| x^{(i)} \|^2} $$

En este punto no estamos determinando $k$ sino definiendo una medida de error. Por ejemplo pensar que la operacion de arriba $< 0.01$. Un error de $1\%$ relativo.

* Suma acumulada relativa de valores singulares   :      

$$\frac{\sum_{i=1}^k  \sigma_i^2}{\sum_{i=1}^m \sigma_i^2} $$

Encontramos ahora una relacion entre las ultimas dos ecuaciones:    


Se puede probar, y esto es un ejercicio:    


$$ \sum_{i=1}^m \| x^{(i)} \|^2  = \sum_{i=1}^m \sigma_i^2  $$


El error total acumulado debido a la proyeccion se debe a la eliminacion de las componentes $k+1, k+2, \cdots, n$. 
 \begin{eqnarray*}
      \begin{pmatrix}
        0 & \cdots & \cdots & \cdots & \cdots & \cdots & \cdots & 0 \\
        \vdots & \ddots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots  \\
        \vdots & \ddots & 0 & \ddots & \vdots & \vdots & \vdots & \vdots  \\
        \vdots & \cdots & 0  & \sigma_{k+1} & 0 & \cdots & \cdots & \vdots \\
        \vdots & \cdots & \cdots & 0 & \sigma_{k+2} &  0 & \cdots & \vdots \\
        \vdots & \ddots & \ddots & \ddots & \ddots & \ddots & \ddots & \vdots \\
        \vdots & \ddots & \ddots & \ddots & \ddots & 0 & \sigma_{n-1} & 0 \\
        0 & \cdots & \cdots & \cdots & \cdots & \cdots & 0 & \sigma_n
      \end{pmatrix} .
    \end{eqnarray*}



$$ \sum_{i=1}^m \| x^{(i)} - x^{(i)}_{approx} \|^2 = \sum_{i=k+1}^n \sigma_i^2 = \sum_{i=1}^n \sigma_i^2 - \sum_{i=1}^k \sigma_i^2 $$

$$ \frac{\epsilon}{\text{Tol}_{\text{var}}} =  \frac{\frac1m \sum_{i=1}^m \| x^{(i)} - x^{(i)}_{\text{approx}} \|^2}{\frac1m \sum_{i=1}^n \| x^{(i)} \|^2}  = \frac{\sum_{i=1}^n \sigma_i^2 - \sum_{i=1}^k \sigma_i^2 }{\sum_{i=1}^m \sigma_i^2} = 1 -  \frac{ \sum_{i=1}^k \sigma_i^2 }{\sum_{i=1}^m \sigma_i^2}$$

Si por ejemplo queremos acumular el 99% de la energia relativa y 1% de error:      

$$ \frac{\frac1m \sum_{i=1}^m \| x^{(i)} - x^{(i)}_{\text{approx}} \|^2}{\frac1m \sum_{i=1}^n \| x^{(i)} \|^2}  = 1 -  \frac{ \sum_{i=1}^k \sigma_i^2 }{\sum_{i=1}^m \sigma_i^2}  \le 0.01  $$

Esto es lo mismo que decir

$$ \frac{ \sum_{i=1}^k \sigma_i^2 }{\sum_{i=1}^m \sigma_i^2}  \ge 0.99$$
