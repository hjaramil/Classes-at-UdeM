**Definición 1.1.5 (Matriz convergenge)**:
Asumimos matrices cuadradas. Decimos que $A$ es convergente si

$$\lim_{n \to \infty} A^n = 0  $$
Divergente, alguna de las componentes va a $\infty$.



**Definición 1.1.6** (radio espectral): matrices cuadradas,
El radio espectral de la matriz $A$ es

$$\rho(A) = \max | \lambda | $$
con $\lambda$ autovalor de $A$.

**Teorema 1.1.1**: La matriz $A$ es convergente si y solo si $\rho(A)< 1$.

"$\implies$"


Entonces

$$A \mathbf{x} = \lambda x  $$
$$A² \mathbf{x} = A (A \mathbf{x} ) = A \lambda \mathbf{x}= \lambda A\mathbf{x} = \lambda² \mathbf{x}  $$
En general
$$A^k  \mathbf{x} = \lambda^k \mathbf{x}  $$

Ahora como $A^k \to 0$, entonces $\lambda^k \mathbf{x} \to 0$
pero $\mathbf{x} \ne 0$, entonces $\lambda^k \to 0$. O sea que
$| \lambda | < 1$. Esto para cuaquier $\lambda$ autovalor de $A$, el maximo tambien tiene que ser que uno. $\max | \lambda | < 1$
$\rho(A) < 1$.

**Teorema 1.1.2**: Para una matriz $A$ (cuadrada)

$$\rho(A) \le \| A \|  $$
no importa que norma.

Veamos. Sea $\mathbf{x}$ no cero. Tal que $A \mathbf{x} = \lambda \mathbf{x}$. Tenemos de la norma de matrices

$$\| A \mathbf{x} \| \le \| A \|  \| \mathbf{x} \| $$
de otro lado $$\| \lambda \mathbf{x} \| = | \lambda | \|  \mathbf{x} \|  $$. De forma que

$$| \lambda | \| \mathbf{x} \|  \le \| A \| \| \mathbf{x} \| $$
Ahora $\| \mathbf{x} \| \ne 0$. Cancelamos

$$| \lambda | \le \|  A \|. $$
Ahora bien $\lambda$ es cualquiera. Entonces el maximo de los $| \lambda |$ sigue siendo mas pequenho que $\| A \|$

$$ \rho(A) \le \| A \|  $$



**Teorema 1.1.3**: La matriz $A$ es invertible si y solo si $\lambda = 0$ no es autovalor de $A$.

Prueba:
Sabemos que $A \mathbf{x} = 0$, para $\mathbf{x} \ne 0$ significa que $A$ no es invertible. De otra forma multiplicamos por $A^{-1}$ y encontramos que $\mathbf{x} = 0$.  
Que $\mathbf{x} \ne 0$, es equivalente a decir que $A \mathbf{x} = 0$, implica que $A \mathbf{x} = \lambda \mathbf{x}$ con $\lambda=0$.




**Definicion 1.1.7 (Disco de Gershgorin)**: Para una matriz $A \in \mathbb{C}^{n \times n}$ con componentes $a_{ij}$, definimos el radio  (me voy por a fila $i$)
$$R_i = \sum_{j \ne i} |a_{ij} |   $$, el **centro** es $a_{ii}.
El disco

$$D(a_{ii}, R_i) = \{ \mathbf{x} \quad |  \quad \| \mathbf{x} - a_{ii}  \| < R_i \}  \subset \mathbb{C} $$
se llama **disco de Gershgorin**.

Ejemplo: Considere la matriz compleja

$$A = \begin{pmatrix}
3 & \mathrm{i}  & 1 \\
-1 & 4 + 5 \mathrm{i} & 2 \\
2 & 1 & -1
  \end{pmatrix}  $$

\begin{eqnarray}
\text{fila 1} &:& \text{centro} \quad (3,0) \quad , \quad \text{radio} = | \mathrm{i} | + | 1 | = 1 + 1 = 2  \\
\text{fila 2} &:& \text{centro} \quad (4,5) \quad , \quad \text{radio} = | -1 | + | 2 | = 3 \\
\text{fila 3} &:& \text{centro} \quad (-1,0) \quad , \quad \text{radio} = |2 | + | 1 | = 3 \\
\end{eqnarray}

**Teorema 1.1.4 (Circulo de Gershgorin)**: Todos los autovalores estan contenidos dentro de los círculos de Gershgorin. Es decir, cada $\lambda$ está en al menos un círculo de Gershgorin de la matriz $A$.


 Prueba. Sea $\lambda$ un autovalor (arbitrario) de $A$.  
 Toemmos el  autovector $\mathbf{x}$ correspondiente a este $\lambda$. Como $\mathbf{x} \ne 0$, entonces existe una componente $i$ tal que $| x_i | \ge | x_j | $, $j=1,2, \cdots, n$. Entonces construimos un vector

 $$\mathbf{y} = \frac{\mathbf{x}}{x_i} $$
 La componente $i$ de $\mathbf{y}$ es $1$. Y las demas estan por debajo de 1.
 $\mathbf{y}$ tambien es autovector.

 Nota: Si $\mathbf{x}$ es autovector, cualquier multiplo $c \mathbf{x}$ tambien lo es:
 Prueba:

 $$A c \mathbf{x} = c A \mathbf{x} = c (\lambda \mathbf{x}) = ( \lambda) c \mathbf{x}  $$

De aca que

$$A \mathbf{y} =  \lambda \mathbf{y}  $$
Tomamos la componente $i$ de este producto (recuerde que $y_i=1$)

$$ \sum_{j} a_{ij} y_j = \lambda y_i = \lambda $$

Ahora re-escribamos esta suma separando la diagonal $a_{ii}$

$$\sum_{j \ne i} a_{ij} y_i + a_{ii} = \lambda $$
(suma sobre la fila $i$)
De aca y la desigualdad triangular

$$| \lambda - a_{ii} | = \left | \sum_{j \ne i} a_{ij} y_j  \right |  \le \sum_{j \ne i} | a_{ij} | | y_j | \le \sum_{j \ne i} | a_{ij} | = R_i . \tag{1} $$

Geometricamente esto me dice que $\lambda$ esta dentro del circulo de Gershgorin con entro en $a_{ii}$.

En el ejemplo anterior, los autovalores son

$$\lambda_1 = 4.11 + 0.72 \mathrm{i} \quad , \quad  \lambda_2 = 3.47 \quad , \quad  \lambda_3 = -1.58 + 0.27 \mathrm{i}  $$
Claramente estan dentro de los circulos de Gershgorin en la Figura. (puntos rojos).

**Comentario**: Asumamos que la matriz $A$ es diagonalmente dominante por filas. Entonces la matriz es invertible, pues cada autovalor satisface $| \lambda - a_{ii} | < R_i$ (ver Ecuacion (1)). Si $\lambda=0$, entonces $|a_{ii}| < R$. Pero
$$R_i = \sum_{i \ne j} | a_{ij}| > | a_{ii} |  $$
entonces por (reduccion al absurdo) $\lambda \ne 0$.
Pero si $\lambda \ne 0$, la matriz es invertible.  (teorema 1.1.3).

Vamos a asumir que la matriz es diagonal. Entonces los $R_i$ son todos 0 y los autovalores son **automaticamente** los valores de la diagonal.  (ver la Ecuacion 1).
Otra forma de ver eso. Es que el polinomio caracteristico es

$$p(\lambda) = \det (A - \lambda I) = \prod_{i=1}^n (a_{ii} - \lambda)  = 0.  $$
Eso es lo mismo que decir que $\lambda = a_{ii}$ son los autovalores.

$$\mathbf{x}_k = T \mathbf{x}_{k-1} + \mathbf{c}  \tag{2} $$

Queremos que el sisema (2) conerja. Entonces para eso se necesita

$$\lim_{n \to \infty} \mathbf{x}_n = \mathbf{x} $$
Esto que quiere decir? (de la Ecuacion (2)

$$\mathbf{x} =  T \mathbf{x} + \mathbf{c}. $$
Esto a su vez quiere decir, (pasando a resar el $ T \mathbf{x}$

$$\mathbf{x} - T \mathbf{x} = \mathbf{c} $$
$$(I - T) \mathbf{x} = \mathbf{c}$$
Es decir

$$\mathbf{x} = (I - T)^{-1} \mathbf{c}. $$
Trataremos de interpretar este resultado.

Primero del algebra, si $T$ fuese una variable algebraica

$$\frac{I - T^n}{I - T} = I + T + T^2 + \cdots + T^{(n-1} $$

Explicacion. Los opeardores tambien tienen propiedades algebraicas, la ley distributiva, pero todo operador conmuta con si mismo y la identidad.

$$I - T^2 = (I - T) ( I + T)  $$
$$I- T^3 = (I-T)(I + T + T^2)  $$
$$ \vdots $$
$$I - T^n = ( I - T)(I + T + \cdots + T^{n-1}) $$
Vamos a asumir que en esta secuencia $T$ converge, es decir que cuando $n \infty$, $\lim_{n \to \infty} T^n = 0$.
Si tomamos este limite veriamos que

$$I  = ( I - T)(I + T + \cdots + T^{n-1} + \cdots )  $$
De aca podemos ver que

$$(I - T)^{-1} \equiv \frac{1}{I - T} = I + T  + T^2 + \cdots + T^n + \cdots  $$

Esto funciona siempre y cuando $T$ converga, $\lim_{n \to \infty}T^n = 0$.

**Teorema 1.1.5.**: Asuma la matriz $T$ de la recursion (2).
El radio espectral de $T$ es $\rho(T) < 1$ si y solo si $(I-T)^{-1}$ existe, ademas

$$(I-T)^{-1} = I + T + T^2 + \cdots = \sum_{i=0}^\infty T^i $$
y esta formula converge.

**Prueba**:

"$\Rightarrow$"

Asumamos que $\rho(T) < 1$ y que $\lambda$ es autovalor de $T$.
Entonces $T \mathbf{x} = \lambda \mathbf{x}$.  Pero significa que

$T \mathbf{x} = \lambda \mathbf{x}$.  
$I - T \mathbf{x} = I - \lambda \mathbf{x}$.  
$(I - T) \mathbf{x} = (1 - \lambda ) \mathbf{x}$.  
De forma que $1 - \lambda$ es autovalor de $I-T$. Como
$\rho(T) < 1$, entonces $| \lambda | < 1$ y $1 - \lambda \ne 0$.
De forma que $I-T$ tiene inversa.

Ahora bien, escribimos


\begin{eqnarray}
\mathbf{x}_i &=& T \mathbf{x}_{i-1} + \mathbf{c} \\
&=& T (T \mathbf{x}_{i-2} + \mathbf{c}) + \mathbf{c} \\
&=& T^2 \mathbf{x}_{i-2} + (T + I) \mathbf{c} \\
&=& T^3 \mathbf{x}_{i-3} + (T^2 + T + I) \mathbf{c} \\
&\vdots&  \\
&=&  T^i \mathbf{x}_{i-i} + (I +  T + \cdots T^{i-1}) \mathbf{c} \\
&=&  T^i \mathbf{x}_{0} + (I +  T + \cdots T^{i-1}) \mathbf{c} \tag{3}.
\end{eqnarray}
Como $\rho(T) < 1$, entonces $\lim_{i \to \infty} T^i \mathbf{x}_0 = 0$.

Tomamos el limite en (3)

$$\mathbf{x} = \lim_{i \to \infty} \mathbf{x}_i = (I + T + T^2 +
\cdots + T^n + \cdots ) \mathbf{c} =  (I - T)^{-1} \mathbf{c}. \tag{4} $$

Es decir,

$$\mathbf{x}_k = T \mathbf{x}_{k-1} + \mathbf{c} $$
tiene solucion $\mathbf{x}$ (4).

"$\Leftarrow$"
Asumimos que $I-T$ es invertible y probamos que $\rho(T)<1$.

Dado que $(I-T)^{-1}=I + T + T^2 + \cdots + T^n + \cdots $, existe el limite $\lim_{n \to \infty} T^n = 0$ (de otra forma la serie diverge).  De acuerdo al teorema 1.1.1, entonces $\rho(T) < 1$.



El ser estrictamente $<1$ es importante. Por ejemplo veamos un caso done $\rho(T)=1$. La matriz identidad $A=I$. Autovalores de la identidad

$$p(\lambda ) = \det ( I - \lambda I) = \det I ( 1- \lambda) = (1 - \lambda)^n $$
La solucion a $p(\lambda) = 0$ es $\lambda=1$. El radio espectral $\rho(I)=1$.
El sistema si $\mathbf{c}=0$. $T=I$

$$\mathbf{x}_k = I \mathbf{x}_{k-1}  \tag{5} $$
$\mathbf{x}_k = \mathbf{x}_{k-1}= \mathbf{x}$.

Caulquier $\mathbf{x}$ satisface esta ecuacion.

La ecuacion (5) tiene todo el espacio como solucion.
Cuando es "<"  la solucion es unica.

Hasta el momento encontramos como resolver el sistema y si tiene o no solucion (unica). Pero el error que?

Vamos a calcular el error de la solucion. Asumimos que la solucion es $\mathbf{x}$. Encontramos la diferencia entre la solucion exacta $\mathbf{x}$  (4) y la aproximada $\mathbf{x}_i$ que es la (3)

\begin{eqnarray}
\mathbf{x} - \mathbf{x}_i &=& (I + T + \cdots + T^{i-1} + T^{i} + \cdots + T^{i+k} + \cdots ) \mathbf{c} - T^i \mathbf{x}_0 - (I + T + \cdots + T^{i-1}) \mathbf{c}   \\
&=& (T^i + T^{i+1} + \cdots + T^{i+k} + \cdots ) \mathbf{c} - T^i \mathbf{x}_0 \\
&=& T^i ( I + T + \cdots + T^k + \cdots) \mathbf{c} - T^i \mathbf{x}_0 \\
&=& T^i \mathbf{x} - T^i \mathbf{x}_0 \\
&=& T^i ( \mathbf{x} - \mathbf{x}_0).
\end{eqnarray}
Como el error se calcula con normas tomamos normas y recuerde $\| A B \| \le \|A \| \| B \|$.

Entonces

$$\| \mathbf{x} - \mathbf{x}_i \| \le \| T \|^i  \| \mathbf{x} - \mathbf{ x_0} \| $$.

Si $\| T \|<1$, entoces $\| T \|^i < 1$ y

* El error es mas pequenho entre mas pequenho sea $\| T \| $.
* El error es mas pequenho cuando $\mathbf{x}$ est'a mas cerca de $\mathbf{x}_0$.

El problema es que no conocemos $\mathbf{x}$.

Vamos a cambiar un poco la estrategia.

\begin{eqnarray}
\| \mathbf{x}_{i+1} - \mathbf{x}_i \| &=& \| T \mathbf{x}_i - T \mathbf{x}_{i-1} \|  \\
&=& \| T ( \mathbf{x}_i - \mathbf{x}_{i-1}) \| \\
&\le& \| T \| \| \mathbf{x}_i - \mathbf{x}_{i-1} \| \\
&\le& \| T \|^2 \| \mathbf{x}_{i-1} - \mathbf{x}_{i-2} \|  \\
&\vdots & \\
&\le& \| T \|^i \| \mathbf{x}_1 - \mathbf{x}_0 \|
\end{eqnarray}

Esto mide el error de una solucion con respecto a la anterior, pero no con respecto a la solucion exacta $\mathbf{x}$.

Para esto necesitamos mas trucos. Existe la ley telescopica de la suma, si $i > j>1$

\begin{eqnarray}
\| \mathbf{x}_i - \mathbf{x}_j  \| &=&
\| \mathbf{x}_i - \mathbf{x}_{i-1} + \mathbf{x}_{i-1}
- \mathbf{x}_{i-2} + \mathbf{x}_{i-2} + \cdots -\mathbf{x}_{j+1} +  \mathbf{x}_{j+1} - \mathbf{x}_j \|  \\
&\le & \| \mathbf{x}_i - \mathbf{x}_{i-1} \|  + \| \mathbf{x}_{i-1} - \mathbf{x}_{i-2} \| + \cdots + \| \mathbf{x}_{j+1} - \mathbf{x}_j \| \\
&=& \left ( \| T \|^{i-1} + \| T \|^{i-2} + \cdots + \| T \|^j)
    \right )  \| \mathbf{x}_1 - \mathbf{x}_0 \|  \\
&=&  \| T \|^j \left ( I + \| T \| + \cdots + \| T \|^{i-j-1}
    \right ) \| \mathbf{x}_1 - \mathbf{x}_0 \| \\
 &=&
 \frac{\| T \|^j ( 1 - \| T \|^{i-j})}{1 - \| T \|} \| \mathbf{x}_1 - \mathbf{x}_0 \|.
\end{eqnarray}
En la ultima ecuacion tomamos el limite cuando $i \to \infty$ entonces
$$\lim_{i \to \infty} \| T \|^{i-j}  = 0 $$
ademas $\lim_{i \to \infty} \mathbf{x}_i = \mathbf{x}$.

Concluimos

$$\| \mathbf{x} - \mathbf{x}_j \| \le  \| T \|^j \frac{\mathbf{x}_1 - \mathbf{x}_0}{1 - \| T \|} \tag{6} $$
