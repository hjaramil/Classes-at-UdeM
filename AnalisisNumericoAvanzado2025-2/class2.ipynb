**Definicion AutoValor-AutoVector**. Sea $A$ una matriz cuadrada (compleja o real) si existe un escalar $\lambda$ y un vector $\bmx$ tales que

$$A \bmx = \lambda \bmx $$
decimos que $\lambda$ es un **autovalor de la matriz $A$ y que $\bmx$ es el
 **autovector** correspondiente a $\lambda$ de la matriz $A$.

Veamos la interpretacion de este concepto, a mano alzada y Wikipedia.


El concepto de autovalor/vector es muy importante en  varias areas del conocimiento:

* Mecanica cuantica. Los estados de energia son autovalres del operador de Schrodringer (la ecuacion de Scrodinger)

* En la teoria de la elasticidad. Los autovalores son frecuencias de estados de vibracion.

* En la optica. El arco-iris es una descomposicion de la luz blanca en el **espectro** de autovalores.  La palabra **autovalor** se asocia con espectro.

* En el dibujo

* En el analisis numerico para encontrar estabilidad de sistemas lineales.

Como se encuentran los autovalores/vectores (tambien les dicen valores propios vectores propios) (en ingles Eigenvalues/Eingenvectors)

$$A \bmx = \lambda \bmx $$
$$A \bmx - \lambda \bmx = 0 $$
$$A \bmx - \lambda I \bmx = 0 $$
$$ ( A  - \lambda I)  \bmx = 0 $$
Como $\bmx \ne 0$ entonces el determinante de la matriz es 0 (algebra lineal)

De aca, algebra lineal que


$$ p(\lambda) = \det( A - \lambda I ) = 0$$
$p(\lambda)$ se llama **polinomio caracteristico**.

**Defincion 1.1.3 (positiva/negativa definida**: Una matriz $A$ es positiva definida si para todo vector $\bmx$ en el espacio ($\mathbb{R}^{n \times n}$)

$$\bmx^T A \bmx > 0  $$
Negativa definida si
$$\bmx^T A \bmx < 0 $$

Se puede probar que si es positiva definida todos autovalores son positivos.

**Definicion 1.1.4 estrictamente diagonal dominante**: La matriz $A$ es estrictamente diagonal dominante por filas, si

$$ | a_{ii} | > \sum_{j=1, j \ne i}  | a_{ij} |   $$

Por columnas si

$$ | a_{jj} | > \sum_{i=1, i \ne j}  | a_{ij} |   $$

Se puede quitar la palabra estricta usando $\ge$.


** Definicion 1.1.5 (Matriz convergente):** La matriz $A$ es convergente si $\lim_{k \to \infty}  A^k = 0$.
Ejemplos en el tablero. (mano alzada)

**Definicion 1.1.6 (radio espectral)**: Al valor

$$ \rho(A) = \max | \lambda| \quad , \quad \text{autovalor de } A $$
se llama **radio espectral**
En palabras, el radio espectral es el mas grande (en valor absoluto) de todos los autovalores.

Todas las matrices en este punto son cuadradas.
**Teorema 1.1.1**: La matriz $A$ es convergente si y solo si $\rho(A) < 1$.

Demostracion: Asumamos primero que $\rho(A) < 1$ por cada autovalor $\lambda_i$, $i=1,2, \cdots, n$ tenemos que $| \lambda_i | < 1$.
Ahora bien, el calculo de la potencia de la matriz $A$ esta dado por.

En el Apendice B de las notas de clase tenemos que


$$A^k = Q J^k Q^{-1} $$
donde $Q$ es una matriz finita (no hay infinitos) y la matriz $J$ tiene muchos ceros y donde no hay ceros hay $\lambda^k$, el cual va para 0 cuando $k \to \infty$ por que $| \lambda | < 1$. Entonces el producto de las tres (que es $A$) va para 0.

Para la otra implicacion (Que si la matriz convege $\rho(A) < 1$)
Asumamos que la matriz converge. Es decir,  por la definicion 1.1.5 tenemos que

$$\lim_{k \infty} A^k = 0 $$
Entonces comoo

$$A \bmx = \lambda \bmx $$
$$A^2 \bmx = A (A \bmx) =  A (\lambda \bmx) = \lambda A \bmx = \lambda \lambda \bmx = \lambda^2 \bmx $$
$$A^3 \bmx = A (A^2 \bmx) =  A (\lambda^2 \bmx) = \lambda^2 A \bmx = \lambda^2 \lambda \bmx = \lambda^3 \bmx $$
$$ \vdots $$

$$A^k \bmx = \lambda^k \bmx $$
Como $A^k$ va para cero, asi tambien va para cero $\lambda^k \bmx$, pero como $\bmx$ es fijo, no cambia con $k$ el que va para cero es $\lambda^k$,

$$\lim_{k \to \infty} \lambda^k = 0$$
En algun momento $\lambda^p < 1$, y por la tanto $\lambda < 1$.
Entonce $\lambda < 1$  y como $\lambda$ es cualquiera (arbitrario) entonces
$\rho(A) < 1$ .

**Teorema 1.1.2** Para una matriz $A$ (cuadrada)

$$\rho(A) \le \| A \|  $$

Asuma que $\bmx \ne 0$ tal que $A \bmx = \lambda \bmx$, $\lambda$ es un autovalor de $A$. Usamos

$$\| A \bmx \| \le \| A \| \| \bmx \| $$
De otro lado $\| \lambda \bmx \| = | \lambda | \| \bmx \|$ de forma que


$$ \| \lambda \bmx \| \le \| A \| \| \bmx \|   $$
$$  | \lambda | \cancel{\|  \bmx \|} \le \| A \| \cancel{\| \bmx \|}   $$
Como $\bmx \ne 0$ , $ \| \bmx \| \ne 0$ y dividimos por la norma de $\bmx$

$|\lambda | \le \| A \|$,

pero $\lambda$ es arbitrario, o sea que **todos** son menores que la norma de $A$ y por esto

$$\rho(A) < \| A \|. $$


**Teorema 1.1.3**: La matriz $A$ es invertible si y solo si $\lambda=0$ **no** es un autovalor de $A$.

Prueba: Sabemos que $A \bmx = 0$, para $\bmx \ne 0$ significa que $A$ no es invertible. Por varias razones:

1. Esto implicaria aque el determinante de $A$ es 0, y entonces $A$ no es invertible.
2. Si $A$ es invertible, multiplique a ambos lados por $A^{-1}$ y
$$ A^{-1}A \bmx = A^{-1}0 = 0    $$
$$  \bmx =  0    $$

Otro argumento
Decir que $A \bmx = 0$ equivale a decir que $A \bmx = \lambda \bmx$, donde
$\bmx$ es autovector de $A$,  es decir que $\lambda \bmx = 0$, pero como $\bmx$ no puede ser cero (por es autovector) entonces $\lambda = 0$.

En palabras, lo que quiero que quede de aca es que si un autovalor (cualquiera)
es 0 entonces la matriz $A$ **no es invertible** (es decir es **singular**).

**Definicion 1.1.7 (Disco de Gershgorin)**: Para una matriz $A \in \mathbb{C}^{n \times n}$ con componentes $a_{ij}$, definimos el radio
$$ R_i = \sum_{j \ne i} | a_{ij}| $$ el centro es $a_{ii}$. El disco
$D(a_{ii}, R) \subset \mathbb{C}$ y radio $R$ se llama el **disco de Gershgorin**.

**Ejemplo 1.1.1**:  Sea $A$ la matriz compleja


$$A = \begin{pmatrix} 3 & \mathrm{i} & 1 \\ -1 & 4 + 5 \mathrm{i} & 2 \\ 2 & 1 & -1  \end{pmatrix}  $$

Los centros y radios fila por fila de la matriz estan dados por

$$\text{fila 1}: \text{centro  } (3,0)  \quad, \quad \text{radio  } |\mathrm{i}| + | 1 | = 2    $$

$$\text{fila 2}: \text{centro  } (4,5)  \quad, \quad \text{radio  } |1| + | 2 | = 3    $$

$$\text{fila 2}: \text{centro  } (-1,0)  \quad, \quad \text{radio  } |2| + | -1 | = 3    $$


Ejemplo 2
$$A = \begin{pmatrix} 3 & \mathrm{i} & 1 \\ -1 & 2 &  4 + \mathrm{i} \\ 2 & 1 & -1  \end{pmatrix}  $$

Los centros y radios fila por fila de la matriz estan dados por

$$\text{fila 1}: \text{centro  } (3,0)  \quad, \quad \text{radio  } |\mathrm{i}| + | 1 | = 2    $$

$$\text{fila 2}: \text{centro  } (2,0)  \quad, \quad \text{radio  } |-1| + | 4 + 5 \mathrm{i}  | =  1 + \sqrt{16 + 25} = 1 + \sqrt{41}    $$

$$\text{fila 2}: \text{centro  } (-1,0)  \quad, \quad \text{radio  } |2| + | -1 | = 3    $$

```python
import numpy as np
A = np.array ( [[3, 1j , 1], [-1, 4+5j , 2], [2, 1, -1 ]])
print(A)
```

```python
# find the eigenvalues of A
w, v = np.linalg.eig(A)
print(w)
```

```python
eigenvalues = np.linalg.eigvals(A)

# Calculate the centers and radii of the Gershgorin circles
centers = np.diag(A)
radii = np.sum(np.abs(A - np.diag(np.diag(A))), axis=1)
```
