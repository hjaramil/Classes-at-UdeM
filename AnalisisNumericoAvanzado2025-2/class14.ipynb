# Diseno del algoritmo
Vamos a asumir que $nx > 3, ny>3$. El numero de filas (columnas) de la matriz $A$ es $N=(nx-1)(ny-1)$ la matriz es de orden $N \times N$ en este caso tiene $(nx-1)^2 (ny-1)^2$ componentes. Son muchas, por eso se debe paralelizar, pero al menos correr sistemas iterativos. Vamos a  usar Gauss-Seidel.

El algoritmo consiste en
* **Llenar la matriz $A$**
    * Diagonal principal. El valor es $\alpha$
    ``A[i,i] = alpha``, $i=0,1, \cdots, N-1$
    * Subdiagonales encima y debajo, inmediatas de la diagonal principal.
    ``A[i+1, i] = A[i, i+1]=1 `` para $i=0,1, \cdots, N-2$
    Ojoooo: Algunos valores de estos son 0. Los reseteamos mas tarde.
    * Las subdiagonales a una distancia $nx-1$ con el valor $\beta^2$.
    Es decir
    ``A[i+nx-1,i] = A[i, i+nx-1] = \beta^2`` $i=0,1, \cdots, N-nx$

    * Identifiquemos que valores del paso 2 en esta sublista son 0.
    En la frontera de la izquierda tenemos los nodos $(0,j)$ el valor $\ell$ aca es $\ell=(j-1)(nx-1) + i= (j-1)(nx-1)$, pues $i=0$,
    entonces ``A[ell, ell-1]=A[ell-1, ell]=0``.

* **Llenar el vector $b$**
    * Capa $j=1$. En general $c_i = -g_i \beta^2$, pero en el caso $i=1$, se le resta $e_1$ y en el caso $i=nx-1$ se le resta $f_1$.
    * Capas $2 \le j \le ny-2$, resta $e_j$ para $i=1$, resta $f_j$ para $i=nx-1$.
    * Ultima capa $j=ny-1$. Se aisgna $c_i = -h_i \beta^2$, pero ojo. Si $i=1$, se resta $e_{ny-1}$ y si  $n=nx-1$, se resta $f_{ny-1}$.
* **Resolver $Ax=b$** : Usamos Gauss-Seidel
* Y convertir el vector $x=w_i$ a una matriz $w_{ij}$ (reshape)
* Se prueba el algoritmo. Para esto disenhamos una solucion analitica. Las soluciones a la Ecuacion de Laplace, se llaman armonicos (en ingles harmonics). Cualquier funcion lineal cumple la ecuacion de Laplace $\nabla^2 f=0$, pues las segundas derivadas de una ecuacion lineal son 0.
Estas no son gracias. Veamos una mas interesante


$$u(x,y) =  \frac{y}{(1 +  x)^2 + y^2} $$
De aca sale facilmente los valores de frontera.

\begin{eqnarray}
u(a,y) = \frac{y}{(1 + a^2)^2 + y^2} = f(y) \\
u(b,y) = \frac{y}{(1 + b^2)^2 + y^2} = e(y) \\
u(x,c) = \frac{c}{(1 + x^2)^2 + c^2} = g(x)  \\
u(x,d) = \frac{d}{(1 + x^2)^2 + d^2} = h(x)
\end{eqnarray}

# Programacion en `Python` de la Ecuacion de Laplace.

```python
# traemos la solucion a Guass-Seidel de la Clase 3.
import numpy as np
from numpy import linalg as LA

def gauss_seidel(A, b, nmax, tol):
    L = np.tril(A) # lower triangular
    U = A-L
    n = len(b)
    x =np.zeros(n) # solucion inicial
    k=0
    error = 10

    LAI = LA.inv(L) # inversa de L

    # Iteraciones de Gauss Seidel
    while( k < nmax and tol < error):
        xnew = np.dot(LAI, b -np.dot(U,x)) # x = -(L+D)^{-1} U x + (L+D)^{-1} b
        error = LA.norm(xnew-x)
        print("solucion para x en iteracion k=%d con error %5.4f"%(k,error))

        print(" ", xnew)
        x = xnew # rotacion
        k +=1

    if k==nmax:
        print("maximo numero de iteraciones fue alcanzado")
        print("probablemente no hay convergencia")

    return x
```

```python
# programamos la ecuacion elipctica (Poisson)
# solve_elliptic
def solve_elliptic(dx,dy, e, f, g, h, a,b,c,d,rho, nmax, tol, verbose):
    # parametros explicados en clase y notas de clase

    # dimensiones en x  y y
    nx = int( (b-a)/dx)
    ny = int( (d-c)/dy)

    # dimension del espacio sololucion
    N = (nx-1)*(ny-1)

    # ejes x  y y
    X = np.linspace(a+dx, b-dx, nx-1)
    Y = np.linspace(c+dy, d-dy, ny-1)

    # inicializacion en ceros  de  la matriz A y el  vector b
    A = np.zeros([N,N])
    b =  np.zeros(N)


    # reportar tamanhos de A,  b
    print("Tama単o de A: ", A.shape)
    print("Tama単o de b: ", b.shape)

    # definicion de variables intermedias
    beta = dx/dy
    alpha = -2.0*(1.0 +  beta**2) # ver notas de clase

    # Lleno de la matriz A
    for i in  range(N):
        # diagonal
        A[i,i] = alpha

        # subdiagonales inmediatas
        if i < N-1:
            A[i+1, i] = 1
            A[i, i+1] = 1

        # subdiagonales remotas
        if i <= N - nx:
            A[i+nx-1,i] = beta**2
            A[i, i+nx-1] = beta**2

    # modificacion de las subdiagonales inmediatas. Algunos valores son 0
    for j in range(2, ny):
        l = (j-1)*(nx-1)
        A[l][l-1]=0.0
        A[l-1][l]=0.0

    # vector b
    for j in range(ny-1):
        for i in range(nx-1):
            l = j*(nx-1)+ i

            # asignacion inicial de densidades rho
            rhol = rho(X[i], Y[j])
            b[l] = dx**2*rhol

            # primera capa
            if j==0:
                b[l] -= g(X[i])*beta**2
                # esquina inferior izquierda
                if i==0:
                    b[l] -= e(Y[j])
                # esquina inferior derecha
                if i==nx-2:
                    b[l] -= f(Y[j])

            # capas intermedias
            if j > 0 and j < ny-2:
                if  i==0:
                    b[l] -= e(Y[j])
                if i==nx-2:
                    b[l] -= f(Y[j])

            # ultima capa
            if j==ny-2:
                b[l] -= h(X[i])*beta**2
                # esquina superior izquierda
                if i==0:
                    b[l] -= e(Y[j])

                # esquina superior derecha
                if i==nx-2:
                    b[l] -= f(Y[j])

    # imprimimos matriz A y vector b
    if verbose:
        print(" matrix A")
        print(A)
        print(" vector B")
        print(b)

    # imprimimos fronteras
    print("Fronteras")
    if verbose:
        print("e")
        [ print( " ", e(Y[j])) for j in range(ny-1)]
        print("f")
        [ print( " ", f(Y[j])) for j in range(ny-1)]
        print("g")
        [ print( " ", g(X[i])) for i in range(nx-1)]
        print("h")
        [ print( " ", h(X[i])) for i in range(nx-1)]


    # resolvemos Ax=b usando Gauss-Seidel
    w = gauss_seidel(A, b, nmax, tol)
    return w
```

```python
# prueba 1
def e(y):
    return y/(1.0 + y**2)

def f(y):
    return  y/(4. + y**2)

def g(x):
    return  0

def h(x):
    return 1./( (1 + x)**2 + 1.)

# fuente ( o densidad rho)
def rho(x,y):
    return 0

# defina gemetria del problema
a = 0
b = 1
c = 0
d = 1
dx = 0.2
dy = 0.2

# para Gauss-Seidel
nmax = 100
tol=0.001

# resolvemos la eliptica
xy = solve_elliptic(dx, dy, e, f, g, h, a,b,c,d,rho, nmax, tol, 1)

```

Convertimos $w$ de vector a matriz (reshape)

```python
nx = int( (b-a)/dx)
ny = int( (d-c)/dy)

# inicilizamos en 0
w =  np.zeros( [nx+1, ny+1])

# llenar la parte interna de la solucion
for j in range(1,ny):
    for i in range(1,nx):
        l = (j-1)*(nx-1)+ i-1
        w[i,j] = xy[l]

# ejes X, Y
X = np.linspace(a,b,nx+1)
Y = np.linspace(c,d, ny+1)

# llenar con condiciones de frontera
# abajo y arriba
for i in range(nx+1):
    w[i,0] = g(X[i])
    w[i,nx] = h(X[i])

# izquierda y derecha
for j in range(ny+1):
    w[0,j] = e(Y[j])
    w[ny,j] = f(Y[j])
```

# Dimensiones superiores.
Estudiamos la Ecuacion de onda en dos dimensiones espaciales y una  temporal.

La ecucion de onda es

$$\nabla^2 u(x,y,t) - \frac{1}{c^2(x,y)} \frac{\partial^2 u(x,y,t)}{\partial t^2} = - g(x,y,t) $$

Podemos asumir condiciones de frontera tipo Dirichlet
$$u(a,y,t)=p(y,t) \quad  , \quad u(b,y,t)=q(y,t) \quad , \quad  u(x,c,t)=r(x,t) \quad , \quad u(x,d,t)=s(x,t)   $$

Condiciones iniciales
$$u(x,y,t_0)= m(x,y) \quad , \quad u_t(x,y,t_0) = n(x,y) $$


Hay muchas formas de discretizar la ecuacion de onda. Aca usamos central en todas partes.  (uso $k$ para el tiempo y como super-indice)

$$
\frac{w_{i+1 j}^k - 2 w_{ij}^k + w_{i-1 j}^k}{\Delta x^2} +
\frac{w_{i j+1}^k - 2 w_{ij}^k + w_{i j-1}^k}{\Delta y^2} +
-\frac{1}{c_{ij}^2} \frac{w_{i j}^{k+1} - 2 w_{ij}^k + w_{i j}^{k-1}}{\Delta y^2}  = - g_{ij}^k
$$



Despejamos el tiempo $k+1$ en terminos de los demas tiempos.
Multipliquemos todo por $-c_{ij}^2  \Delta t^2$

$$
w_{ij}^{k+1} - 2 w_{ij}^k + w_{ij}^{k-1} =
\frac{c_{ij}^2 \Delta t^2}{\Delta x^2}  (w_{i+1  j}^k - 2 w_{ij}^k + w_{i-1 j}^k ) +
\frac{c_{ij}^2 \Delta t^2}{\Delta y^2}  (w_{i  j+1}^k - 2 w_{ij}^k + w_{i j-1}^k )  + c_{ij}^2 \Delta t^2 g_{ij}^k
$$

De donde
$$
w_{ij}^{k+1}=
\frac{c_{ij}^2 \Delta t^2}{\Delta x^2}  (w_{i+1  j}^k - 2 w_{ij}^k + w_{i-1 j}^k ) +
\frac{c_{ij}^2 \Delta t^2}{\Delta y^2}  (w_{i  j+1}^k - 2 w_{ij}^k + w_{i j-1}^k )  + c_{ij}^2 \Delta t^2 g_{ij}^k + 2 w_{ij}^k - w_{ij}^{k-1}
$$

Re-agrupando. Recojemos terminos en $w_{ij}^k$,
ademas llamando
$\alpha=\frac{c_{ij} \Delta t}{\Delta x}$
$\beta=\frac{c_{ij} \Delta t}{\Delta y}$

$$w_{ij}^{k+1} = \alpha_{ij}^2 ( w_{i+1 j}^k + w_{i-1 j}^k) + \beta_{ij}^2 (w_{i j+1}^k - w_{i j-1}^k) + 2(1 - \alpha_{ij}^2 - \beta_{ij}^2) w_{ij}^k - w_{ij}^{k-1} + c_{ij}^2 \Delta t^2 g_{ij}^k   \tag{1} $$

## Analisis de estabilidad (metodo de Jon Von Neumann)

Para esto asumimos, usando separacion de variables, la solucion de la forma

$$w_{ij}^k =
\mathrm{e}^{\mathrm{i} (i k_x \Delta x + j k_y \Delta y + k \omega \Delta t)}  \tag{2} $$

El factor de ganancia es


$$G = \frac{w_{ij}^{k+1}}{w_{ij}^k}  =
\frac{\mathrm{e}^{\mathrm{i} (i k_x \Delta x + j k_y \Delta y + (k+1) \omega \Delta t)}}{  
\mathrm{e}^{\mathrm{i} (i k_x \Delta x + j k_y \Delta y + k \omega \Delta t)} }= \mathrm{e}^{\mathrm{i} \omega \Delta t}$$

Usando la Ecuacion (1), reemplazando $w_{ij}^k$ en la Ecuacion (2) y dividiendo por $w_{ij}^k y con $g_{ij}^k=0$ (sin fuente). No se necesita fuente para analsis de estabilidad
obtenemos

$$G= 2 \alpha^2 \cos (k_x \Delta x) + 2 \beta^2 \cos(k_y \Delta y) + 2(1 - \alpha^2 - \beta^2) - G^{-1}$$



Multiplicando todo por $G$ y reorganizando terminos

$$G^2 - \phi G + 1 = 0 \tag{3}  $$
donde

$$\phi = 2 \alpha^2 \cos (k_x \Delta x)  + 2 \beta^2 \cos ( k_y \Delta y) + 2(1 - \alpha^2 - \beta^2 )  $$

Como hemos hecho varias veces usando el coseno del angulo medio

$$2 \alpha^2 ( \cos (k_x \Delta x) -1) = 2 \alpha^2 \left ( 1 - 2 \sin^2 \frac{k_x \Delta x}{2} - 1 \right ) = - 4 \sin^2 \left ( \frac{k_x \Delta x}{2} \right ) $$

De igual forma, cambiando $x$ por $y$ y $\alpha$ por $\beta$
$$2 \beta^2 ( \cos (k_y \Delta y) -1) = 2 \beta^2 \left ( 1 - 2 \sin^2 \frac{k_y \Delta y}{2} - 1 \right ) = - 4 \sin^2 \left ( \frac{k_y \Delta y}{2} \right ) $$

Podemos escribir

$$\phi =  - 4 \sin^2 \left ( \frac{k_x \Delta x}{2} \right ) - 4 \sin^2 \left ( \frac{k_y \Delta y}{2} \right ) + 2 = 2 \psi$$

Donde

$$\psi = - 2 \sin^2 \left ( \frac{k_x \Delta x}{2} \right ) - 2 \sin^2 \left ( \frac{k_y \Delta y}{2} \right ) + 1 $$

Resolviendo la Cuadratica (3)


$$G =
\frac{\phi \pm \sqrt{\phi^2 -4}}{2}=
\frac{2 \psi \pm \sqrt{4 \psi^2 -4}}{2}= \psi \pm \sqrt{\psi^2 - 1}
  $$


Veamos  a ver casos.
* $| \psi | >1$:  es posible que $| \psi \pm \sqrt{\psi^2 -1 }  > 1  |$
considere los dos casos $\psi >0$ (con el " + ") y $\psi < 0$ con el "-". Etonces  $|G| >1$.  

* Necesitamos que $| \psi | \le 1$. En cuyo caso
como $G= \psi \pm \mathrm{i} \sqrt{1 - \psi^2}  $


$$| G |^2 = G G^* = \psi^2 + (1 - \psi^2) = 1 $$
Lo que hace que el sistema  se  estable.

De donde para la estabilidad necesitamos
$| \psi | \le 1$.

Es decir
$$
|  - 2 \alpha^2 \sin^2 \left ( \frac{k_x \Delta x}{2} \right ) - 2 \beta^2 \sin^2 \left ( \frac{k_y \Delta y}{2} \right ) + 1 | \le 1
$$

De la identidad $|x | \le 1  -1 \le x \le 1$

o sea que

$$-1 \le  - 2 \alpha^2 \sin^2 \left ( \frac{k_x \Delta x}{2} \right ) - 2 \beta^2 \sin^2 \left ( \frac{k_y \Delta y}{2} \right ) + 1 \le 1 $$

restamos 1  en todas partes.

$$-2 \le  - 2 \alpha^2 \sin^2 \left ( \frac{k_x \Delta x}{2} \right ) - 2 \beta^2 \sin^2 \left ( \frac{k_y \Delta y}{2} \right )  \le 0  $$

Es decir

$$2 >    2 \alpha^2 \sin^2 \left ( \frac{k_x \Delta x}{2} \right ) + 2 \beta^2   \sin^2 \left ( \frac{k_y \Delta y}{2} \right )  $$

De forma que

$$\alpha^2 \sin^2 ( k_x \Delta x/2) + \beta^2 \sin^2 (k_y \Delta y/2) \e 1 $$

Para obtener esto podemos decir que
$$\alpha < 1/\sqrt{2} $$
$$\beta < 1/\sqrt{2} $$
como $\alpha=c \Delta t/\Delta x
tenemos que

$$\Delta t^2 \left(
    \frac{c^2}{\Delta x^2} +
    \frac{c^2}{\Delta y^2}
    \right ) \le 1$$

o de otra forma

$$\Delta t \le \frac{1}{c \sqrt{ \frac{1}{\Delta x^2} + \frac{1}{\Delta y^2}}} $$

Esta es la condicon de estabilidad. Si vamos a 1D,  obtenemos la misma ecuacion donde no escriba el termino en $\Delta y$. Es la condicon CFL que vimos antes

$$\Delta  t \le \max_{c_{ij}} \frac{1}{c_{ij} \sqrt{\frac{1}{\Delta x^2} + \frac{1}{\Delta y^2}}} $$
