# Analisis de estabilidad en la ecuación de Onda: Número CFL.
Estamos resolviendo la Ecuación de onda:

$$\frac{\partial^2 u}{\partial x^2} - \frac{1}{c^2(x)} \frac{\partial^2 u}{\partial t^2} = - g(x,t) \tag{0}$$

Asumimos una cuerda en el intervalo $[a,b]$
* **condiciones de frontera**
$$u(a,t) = e(t) \quad , \quad u(b,t) = f(t) $$
* **condiciones iniciales**
$$u(x,t_0) =  p(x) \quad , \quad u_t(x,t_0) = q(x) $$

En esta ecuacion $c=c(x)$ es la velocidad de propagaci'on de la onda. Puede ser variable.

Vamos a usar para empezar el sistema de derivadas parciales centradas tanto en tiempo como en espacio.

Recuerde
$[a,b]$ se parte en $x_0=a, x_1=a + \Delta x, \cdots x_i=a+i \Delta x$ etc.
$t_j=t_0 + j \Delta t$ para $g$
$$g_{ij} = g(x_i, t_j) $$


$$w_{i \; j+1} = \gamma_i^2 w_{i-1 \; j} + 2(1 - \gamma_i^2)w_{ij}  + \gamma_i^2 w_{i+1 \; j}   - w_{i j-1} + g_{ij} c_i^2 \Delta t^2 \tag{2}$$


Recuerde la ganancia

$$G=\frac{w_{i \; j+1}}{w_{ij}} \le 1 $$

Recuerde tambien del metodo de separacion de variables

$$u_{\omega k}(x,t) = C(\omega, k) \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x}  \tag{3}$$

Podemos asumir $C(\omega,k)=1$ para efectors de analisis de estabilidad.

### Sobre la relaci'on de dispersion.
Si asumimos la solucion
entonces

$$\frac{\partial^2 u}{\partial x^2} =-C(\omega, k) k^2  \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x} $$
De la misma forma

$$\frac{\partial^2 u}{\partial t^2}= -(\omega, k) \omega^2 \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x}. $$

Si ponenmos esto en la Ecuacion de onda con fuente =0

$$
-C(\omega, k) k^2  \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x}
+C(\omega, k) \frac{1}{c^2(x)} (\omega, k) \omega^2 \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x}  = 0
$$


Cancelando
$$ \frac{\omega^2}{c^2(x)}- k^2 = 0. $$

A esta relaci'on se le conoce como la **dispersion relation** (relacion de dispersion. Lo que indica que $k, \omega$ no son independientes y podemos escribir

$$ k = \pm \frac{\omega}{c(x)}  $$

Sigamos con el analisis de estabilidad.

$$w_{ij} = \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x} $$
Cuando insertamos esto en la Ecuaci'on (2)

Asumimos velocidad constante y $\gamma_i=\gamma$ y $g=0$.

$$
\mathrm{e}^{\mathrm{i} (\omega t_{j+1} + k x_i)} = \gamma^2 \mathrm{e}^{\omega t_j + k x_{i-1}} + 2(1 - \gamma^2) \mathrm{e}^{\mathrm{i}( \omega t_j + k x_i)}  \gamma^2 \mathrm{e}^{\mathrm{i}(\omega t_j + k x_{i+1})}  -
\mathrm{e}^{\mathrm{i}(\omega t_{j-1} + k x_j)}
$$

Dividimos todo por $w_{ij}$ para obtener la ganancia. Obtenemos

$$G = \mathrm{e}^{\mathrm{i} \omega \Delta t} = \gamma^2 \mathrm{e}^{-\mathrm{i} k \Delta x} + 2(1 - \gamma^2) + \gamma^2 \mathrm{e}^{\mathrm{i} k \Delta x} - \mathrm{e}^{-\mathrm{i} \omega \Delta t} $$

$$G = \mathrm{e}^{\mathrm{i} \omega \Delta t} = \gamma^2 \mathrm{e}^{-\mathrm{i} k \Delta x} + 2(1 - \gamma^2) + \gamma^2 \mathrm{e}^{\mathrm{i} k \Delta x} - G^{-1} = 2 \gamma^2 \cos k \Delta x + (1 - \gamma^2) - G^{-1}$$

Multiplicamos por $G$ para eliminar el $G^{-1}$

$$G^2 = 2(1 - \gamma^2) G + 2\gamma^2 G \cos ( k \Delta x) - 1 \tag{4} $$

Miremos el coeficiente de $G$
\begin{eqnarray}
2(1 - \gamma^2) + 2 \gamma^2 \cos ( k \Delta x) &=& 2 \left [
(1 - \gamma^2) + \gamma^2 \cos k \Delta x    \right ] \\
&=& 2 \left [ 1 - \gamma^2 + \gamma^2 \left ( 1 - 2 \sin^2 \frac{k \Delta x}{2} \right )  
    \right ] \\
    &=& 2 \left [ 1 - 2 \gamma^2 \sin^2 \left ( \frac{k \Delta x}{2}  \right )  \right ]
\end{eqnarray}


Definimos la variable intermedia

$$\beta= 1 - 2 \gamma^2 \sin^2 \left ( \frac{k \Delta x}{2}  \right )  $$

Entonces tenemos de la Ecuacion (4)

$$G^2 - 2 \beta G + 1 = 0. $$

Si resolvemos esta cuadratica

$$G_{1,2} = \frac{2 \beta \pm \sqrt{4 \beta^2 - 4}}{2} = \beta \pm \sqrt{\beta^2 - 1} $$

Recuerde que queremos $|G| \le 1$. Si $|\beta|\ge 1$ entonces una de las dos soluciones es mayor o igual a 1

Necesitams que $| \beta | \le 1$.
En este caso podemos escribir las dos soluciones para $G$ como

$$G_{1,2} = \beta \pm \mathrm{i} \sqrt{1 - \beta^2} $$

Hallemos la magnitud de $G$

$$|G|^2 = G G^* = \beta^2 + (1 - \beta^2) = 1 $$
O sea que la maginitud de $G$ es 1 y el sistema es estable.

Es decir, necesitamos $| \beta | \le 1$
en otras palabras

$$
\left |
1 - 2 \gamma^2 \sin \left ( \frac{k \Delta x}{2}  \right )
\right | \le 1
$$

Podemos cambiar el signo dentro de las barras y tomar la desigualda. Sumamos 1 a ambos lados y tenemos

$$2 \gamma^2 \sin^2 \left (  \frac{k \Delta x}{2} \right ) \le 2$$
$$ \gamma^2 \sin^2 \left (  \frac{k \Delta x}{2} \right ) \le 1 $$

Como $\sin^2 (k \Delta x/2) \le 1$ necesitamos que
$$\gamma^2 \le 1 $$
Es decir

$$ \gamma \le 1$$.
$\gamma = c \Delta t/\Delta x$

Entonces la estabilidad ocurre cuando

$$\frac{c \Delta t}{\Delta x} \le 1 $$

A esta condici'on se le conoce con el nombre de
[CFL](https://en.wikipedia.org/wiki/Courant%E2%80%93Friedrichs%E2%80%93Lewy_condition) de Courant Friedrichs Lewis.

Como interpretamos este numero?

image.png

Si queremos que $c \Delta t/\Delta x \le 1$

$$c \le \frac{\Delta x}{\Delta t} $$

Es decir

$$\Delta t \le \frac{\Delta x}{c} $$

Vamos a navegar por `Python` un rato.
$$\frac{\partial^2 u}{\partial x^2} - \frac{1}{c^2(x)} \frac{\partial^2 u}{\partial t^2} = - g(x,t) \tag{0}$$

Asumimos una cuerda en el intervalo $[a,b]$
* **condiciones de frontera**
$$u(a,t) = e(t) \quad , \quad u(b,t) = f(t) $$
* **condiciones iniciales**
$$u(x,t_0) =  p(x) \quad , \quad u_t(x,t_0) = q(x) $$

En este caso

$$p(x)= 6 \sin \left ( \frac{\pi x}{L} \right )  $$

La solucion por el metodo de separacion de variables es:

$$u(x,t) = 6 \sin \left ( \frac{\pi x}{L}  \right ) \cos \left ( \frac{\pi c t}{L} \right ) .  $$

Veamos esta solucion (animacion) en Python.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.pyplot import figure

# parameters
L=1.0
c=1.0
A=6.0  # amplitud
nx=200
x= np.linspace(0, L, nx)

# solution of wave equation
def u(x,t):
    return A*np.sin(np.pi*x/L)*np.cos(np.pi*c*t/L)

# figure size and axis
fig, ax = plt.subplots(figsize=(10,6))

line, = ax.plot([], [], lw=2)

ax.set_xlim((0, L))
ax.set_ylim((-A, A))
ax.set_xlabel('x')
ax.set_ylabel('u(x,t)')
ax.set_title('Wave equation')

# initialization function
def init():
    line.set_data([], [])
    return line,

# animation
def animate(frame):
    t = frame*0.02
    y = u(x,t)
    line.set_data(x, y)
    ax.set_title(f"Wave at t={t:.2f}")
    return line,
# creamos la animacion
ani = FuncAnimation(fig, animate, init_func=init, frames=100, interval=20, blit=True)

# wave the  animation
# ani.save('wave.gif', writer='Pillow', fps=30)

# show the animation
plt.show()
```

```python
# implementacion numerica
def solve_waveeq(c, a, b, t0, nt, nx, dt, e, f , p, q , g):

    # calculamos la velocidad maxima
    X = np.linspace(a, b, nx)
    cmax = max(c(xx) for xx in X)

    print("maxima velocidad", cmax)

    # compute L, dx, gamma, stab_cond
    L=b-a
    dx = L/(nx-1)
    gamma=cmax*dt/dx

    # verificamos condicion de estabilidad
    stab_cond=dx/cmax
    if (dt >= stab_cond):
        print("La condicion de estabilidad no se cumple")
        print(f" dt={dt} debe ser menor que dx/c={dx/cmax} ")
        # return

    tmax = t0 + nt*dt
    t=np.arange(t0, tmax, dt)

    # inicializar w
    w = np.zeros((nx,nt), dtype=float)

    print("dimensiones de w ", w.shape)

    # condiciones iniciales
    for i in range(1, nx):
        w[i,0] = p(X[i])

    # falta la condicion en la derivada. Arreglar
    # condiciones de frontera
    for j in range(nt):
        w[0,j] = e(t[j])
        w[nx-1,j] = f(t[j])

    # solucion numerica
    # llenar la primera capa
    for i in range(1, nx-1):
        gamma = c(X[i])*dt/dx
        w[i,1]= (0.5*gamma**2*p(X[i-1]) + (1.0 - gamma**2)*p(X[i]) +
            0.5*gamma**2*p(X[i+1]) + q(X[i])*dt
            + 0.5*c(X[i])**2*dt**2*g(X[i],  t0))

    # como ya tenemos dos capas comenzamos las iteraciones en tiempo
    for j in range(1, nt-1):
        for i in range(1, nx-1):
            gamma = c(X[i])*dt/dx
            w[i,  j+1] = (gamma**2 * w[i-1, j] + 2.*(1 - gamma**2)*w[i,j]
            + gamma**2*w[i+1,j] - w[i,j-1]
            + g(X[i], t[j])*c(X[i])**2*dt**2)

    return w
```

```python
# probamos el algoritmo
def v(x):
    return 1

# condiciones de bordo e iniciales
def e(x):
    return 0

def f(x):
    return 0

def p(x):
    return 6*np.sin(np.pi*x/L)

def q(x):
    return 0

# fuente 0
def g(x,t):
    return 0

dt = 0.001
nt=1001
nx=101
a=0
b=1
t0=0

w = solve_waveeq(v, a, b, t0, nt, nx, dt, e, f, p, q, g)
```

```python
# graficamos columnas de w
params={'legend.fontsize':5, 'legend.handlelength':1}

figure(figsize=(10,6), dpi=80)

Ndt = 110

# labels
plt.xlabel(r"$x$")
plt.ylabel(r"desplazamiento")

# title
plt.title("curvas de onda")

X = np.linspace(a,b,w.shape[0])
for j in range(0, nt, Ndt):
    t=t0+j*dt
    plt.plot(X, w[:,j], label=f"t={t:.2f}")
    plt.legend(labelspacing=1, title=" tiempo")

plt.grid()
plt.show()
```

```python
# graficar error
figure(figsize=(16,10))

nt=1001
dt=0.001
Nt=11
dx=(b-a)/(nx-1)
error=np.zeros((nx, Nt+1))

nskip=100
k=0
for j in range(0, nt, nskip):
    t = j*dt
    for i in range(1, nx-1):
        x=i*dx
        error[i,k]=w[i,j]-u(x,t)
    k+=1


# grafique el error
params={'legend.fontsize':5, 'legend.handlength':1}
plt.xlabel(r"$x$")
plt.ylabel(r'$error$')
plt.title(r'Curvas de error. Solución Numérica')

X = np.linspace(0,L,w.shape[0])
Dt=0.1
for i in range(0, Nt):
    t = i*Dt
    # print("t=",t)
    plt.plot(X, error[:,i], label=str(round(t,1))+ ' seg')
    plt.grid(True)
    plt.legend(labelspacing=1, title=" tiempo")

plt.grid()
plt.show()
```

## Como probar codigo.
* Se puede inventar una solucion con condiciones iniciales y de frontera.
* Se puede generar con esta soluci'on una fuente artificial $g(x,t)$
Resolver esta ecuacion de onda con fuente a la cual se le conoce la solucion.

Por ejemplo queremos condiciones de frontera 0 en $[0,L]$

$$u(x,t) = x(L-x) \sin t = xL \sin t - x^2 \sin t $$
condiciones de frontera
$$u(0,t) = 0 \quad , \quad u(L,t)=0 $$

condiciones iniciales
$$t=0 \quad u(x,0)=0 $$

$$t=0 \quad \partial u/\partial t(x,0)=x(L-x) $$
Con esto creamos la ecuacion de onda con fuente.


Veamos

$$u_x = L \sin t - 2x \sin t \quad , \quad u_{xx} = -2 \sin t$$

$$u_t =  x(L-x) \cos t  \quad , \quad -x(L-x) \sin t$$

Aplicamos esto a la ecuacion de onda.

\begin{eqnarray}
u_{xx} - \frac{1}{c^2} u_{tt} =  -2 \sin t + \frac{1}{c_2} x(L-x) \sin t = g(x,t)
\end{eqnarray}
