# Ecuaciones diferenciales parciales de segundo orden (second order PDEs)

La ecuacion general parcial de segundo orden es de la forma

$$F(x_1, x_2, \cdots, x_n, u, u_{x_1}, u_{x_2} \cdots, u_{x_n}, u_{x_1 x_1}, u_{x_1 x_2}, \cdots u_{x_n x_n}) = 0 .$$


Otra forma
$$F(x_1, x_2, \cdots, x_n, u, u_{,1}, u_{,2} \cdots, u_{,n}, u_{,11},
u_{x,12 }, \cdots u_{,n n}) = 0 .$$
donde

$$u_{,ij} = \frac{\partial u}{\partial x_i \partial x_j} \quad , \quad i=1,2, \cdots, n \quad , \quad j=1,2, \cdots, n .$$

Por ejemplo

$$ L[u] = a u_{xx} + 2 b u_{xy} + c u_{yy} + d u_x + e u_y + fu = g .$$
La razon del 2 es:
* Asumimos que $u_{xy}=u_{yx}$ y se suman en $2$.
* El discriminante (lo vemos en un minuto) se "simplifica" si usamos el 2.

El discriminante de este sistema es

$D = (2 b)^2 - 4 ac = 4 (b^2 - ac) $
pero el 4 se ignora. O sea que
$D = b^2 - ac $

El discriminante sirve para **clasificar** las ecuaciones como sigue

1. **Elipticas**: $D < 0$, por ejemplo, la ecuacion de Laplace
$$ u_{xx} + u_{yy} = 0.$$

2. **Parabolicas**: $D=0$, por ejemplo la ecuacion de calor.
$u_t = \alpha u_{xx}

3. **Hiperbolicas**: $D>0$. Por ejemplo la ecuacion de onda

$$u_{tt} = \alpha u_{xx}$$

## Ecuaciones Hiperbolicas.
### Coeficientes constantes.
Asumimos

$$Lu = a u_{tt} + 2 b u_{tx} + c u_{xx} = 0 \quad , x \in \mathbb{R}, t> 0 .$$
$D>0$, $b^2 > ac \tag{11} $.
Pensamos en condiciones iniciales IVP:


\begin{eqnarray}
u(x,0) &=& \varphi(x) \\
u_t(x,0) &=& \psi(x)
\end{eqnarray}
Para facilitar el analisis, vamos  llamar

$$T = \frac{\partial }{\partial t}  \quad , \quad
X = \frac{\partial }{\partial x} .$$
Entonces la Ecuacion (11) se convierte en

$$ a T^2 + 2 b TX + c X^2 = 0. \tag{12} $$
Como esto ya es algebra, esta ecuacion tiene dos raices (piense en que $T$ es la incognita)

$$\lambda_{1,2} = \frac{-2 b X \pm \sqrt{4 b^2 X^2 - 4 ac X^2}}{2 a}
 =\frac{-b \pm \sqrt{b^2 - ac}}{a} X.$$

De esta forma podemos factorizar (12) como

$$
\left ( \frac{\partial }{\partial t} - \lambda_1 \frac{\partial}{\partial x} \right)
\left ( \frac{\partial }{\partial t} - \lambda_2 \frac{\partial}{\partial x} \right)  = 0.
$$

Esto da lugar dos ecuaciones diferenciales parciales

$$ \left ( \frac{\partial }{\partial t} - \lambda_1 \frac{\partial}{\partial x} \right)  = 0 $$

$$ \left ( \frac{\partial }{\partial t} - \lambda_2 \frac{\partial}{\partial x} \right)  = 0 $$


Usanso el metodo de las caracteristicas vemos que


$$\frac{dx}{dt} = \lambda_1 \quad , \quad \frac{dx}{dt} = \lambda_2 .$$

Si resolvemeos estas dos ecuaciones ($\lambda_i, i=1,2$ son constantes).

$$x = \lambda_1 t + k_1 \quad \quad x= \lambda_2 t + k_2$$

image.png

Redefinimos

$$ \xi = x - \lambda_1 t  \quad , \quad \eta = x - \lambda_2 t .$$
De esta forma la solucion general esta dada por


$$u(x,y) = f(\xi) + g(\eta) .$$
donde $f,g$ son funciones arbitrarias que se pueden determinar a partir de las condiciones iniciales y de frontera.

**Interpretacion fisica**:
* $f(x - \lambda_1 t)$ es una onda moviendese en la direccion de $\lambda_1$.
* $f(x - \lambda_2 t)$ es una onda moviendese en la direccion de $\lambda_1$.

Si insertamos $\xi, \eta$ de forma que ahora $x=x(\xi, \eta), t=t(\xi, \eta)$ entonces en las nuevas variables $\xi, \eta$ encontramos la nueva ecuacion

$$ u_{\xi \eta}  = \frac{\partial^2 u}{\partial \xi \partial \eta} = \frac{\
\partial}{\partial \xi}  \left ( \frac{\partial}{\partial \eta }    \right ) u  = 0 .$$
A esto se le llama **la forma canonica de la hiperbolica**.

Esta forma es mas facil de integrar. Veamos

Definamos

$$ u_{\eta} = v( \xi, \eta) \tag{1} .$$

De forma que tenemos

$$ \frac{\partial v}{\partial \xi } = \left . \frac{dv}{d \xi} \right |_{\eta=\text{const}} = 0  $$
Integrando esta

$$v(\xi, \eta) = g(\eta) .$$
Tenemos

Usando (1)
$$ \left . \frac{d}{d \eta} u \right |_{\xi= \text{const}} = g(\eta)  .$$

Como necesito $u$, integro

$$ u \left .  \right |_{\xi = \text{const}}  = \int g(\eta) d \eta + f(\xi)$$
donde $f$ es una nueva funcion que solo depende de $\xi$.

De forma que podemos escribir

$$u(x,t) = f(\xi) + g(\eta) = f(x - \lambda_1 t) + g(x - \lambda_1 t) .$$
donde redefinimos $g(\eta) \leftarrow \int g(\eta) d \eta $




Como encontramos $f,g$?

## Datos iniciales (IVP).
Vamos a asumir las siguientes condiciones iniciales.
$$ \varphi(x) = u(x,0) \quad , \quad \psi(x) = u_t(x,0) \tag{1a} $$

Etonces tenemos que

$$u(x,0) = f(x) + g(x) = \varphi(x)  \implies f(x) = \varphi(x) - g(x)  \tag{2}$$
$$ \frac{\partial u(x,0)}{\partial t} = \lambda_1 f'(x) + \lambda_2 g'(x) = -\psi(x) $$


Integramos esta ecuacion en $0$ y $x$

$$\lambda_1 f(x) + \lambda_2 g(x) = - \int_0^x \psi(s) ds .$$
Insertamos (2) en esta esta ecuacion


\begin{eqnarray}
\lambda_1 ( \varphi(x) - g(x)) + \lambda_2 g(x) &=& -\int_0^x \psi(x) ds  \\
\end{eqnarray}

\begin{eqnarray}
g(x) = - \frac{1}{\lambda_2 - \lambda_2} \int_0^x \psi(s) ds - \frac{\lambda_1}{\lambda_2 - \lambda_1} \varphi(x) \\
\end{eqnarray}


De forma que
\begin{eqnarray}
f(x) &=& \varphi(x) - g(x) \\
&=& \left (  1 + \frac{\lambda_1}{\lambda_2 - \lambda_1} \right ) \varphi(x)   \frac{1}{\lambda_2 - \lambda_1} \int_0^x \psi(s) ds \\
&=& \frac{\lambda_2}{\lambda_2 - \lambda_1} \varphi(x) + \frac{1}{\lambda_2 - \lambda_1} \int_0^x \psi(s) ds.
\end{eqnarray}
Ya tenemos $f,g$ en funcion de $\varphi, \psi$. Podemos encontrar ahora la solucion $u(x,t)$ usando las condiciones iniciales.


$$u(x,t) = f(\xi) + g(\eta) = \frac{\lambda_2 \varphi(x - \lambda_1 t ) - \lambda_1 \varphi(x - \lambda_2 t)}{\lambda_2 - \lambda_1}  - \frac{1}{\lambda_2 - \lambda_1} \int_{x - \lambda_1 t}^{x - \lambda_2t} \psi(s) ds.$$


A esto se le llama d'Alembert solution.

La linea verde en la figura se llama **dominio de dependencia** (doomain of dependence).

Veamos un ejemplo

**Ejemplo 1.**:
Considere la siguiente ecuacion (IVP)

$$u_{tt}(x,t) = c^2 u_{xx}  \quad -\infty < x < \infty .$$
con las condiciones iniciales (1a).

Solucion: La ecuacion caracteristica es:

$$\lambda^2 - c^2 = 0  .$$
de forma que $\lambda_1= -c , \lambda_2=c$

La ecuacion de d'Alembert dice

\begin{eqnarray} u(x,t) &=& \frac{\lambda_2 \varphi(x - \lambda_1 t ) - \lambda_1 \varphi(x - \lambda_2 t)}{\lambda_2 - \lambda_1}  - \frac{1}{\lambda_2 - \lambda_1} \int_{x - \lambda_1 t}^{x - \lambda_2t} \psi(s) ds \\
&=& \frac{c \varphi(x + c t)  + c \varphi(x -c t)}{2 c} - \frac{1}{2c} \int_0^x \psi(s) ds.
\end{eqnarray}
De forma que
$$ u(x,t) = \frac{\varphi(x+ ct) + \varphi(x - ct)}{2} - \frac{1}{2c} \int_0^x \varphi(s) ds.$$

El problema general con **coeficientes constantes** (el cambio a coeficientes variables es poco)

$$a u_{xx} + b u_{xy} + c u_{yy} + d u_x + e u_y + f u + g = 0 .$$

Este problema se puede simplifcar llevandolo a la **forma canonica**.
Parametrizar $\xi=\xi(x,y), \eta=\eta(x,y)$, de forma que tenemos


$$u(\xi, \eta) = u( \xi(x,y), \eta(x,y)) .$$
En estas nuevas variables aparece una nueva ecuacion

$$a^* u_{\xi \xi} + b^* u_{\xi \eta} + c^* u_{\eta \eta} + d^* u_\xi + e^* u_\eta + f^* u + g^* = 0 .$$

donde (esto es tarea voluntaria)

\begin{eqnarray}
a^* &=& a \xi_x^2 + 2 b \xi_x \xi_y + c \xi_y^2 \\
b^* &=& a \xi_x \eta_x + b ( \xi_x \eta_y + \xi_y \eta_x) + c \xi_y \eta_y \\
c^* &=& a \eta_x^2 + 2 b \eta_x \eta_y + c \eta_y^2 \\
d^* &=& d \xi_x + e \xi_y + a \xi_{xx} + 2 b \xi_{xy} + c \xi_{yy} \\
f^* &=& f \\
g^* &=& g \\
\end{eqnarray}

Observamos que sdi $\lambda_1, \lambda_2$ son las raices de la cuadratica
$a x^2 + 2 bx + c = 0$, entonces

$$\xi = y - \lambda_1 x \quad , \quad \eta = y - \lambda_2 x .$$

Con estos $\xi, \eta$, la ecuacion se simplifica, $a^*=0=c^*$ y la ecuacion queda

$$ b^* u_{\xi \eta} + d^* u_\xi + e^* u_{\eta} + f^* u + g^* = 0 .$$
Pedimos verificar esto.

## Coeficientes variables en la ecuacion hiperbolica.

$$a(x,t) u_{tt} + 2 b(x,t) u_{tx} + c(x,t) u_{xx} + d(x,t) u_t + e(x,t) u_x + f(x,t) u = g(x,t)  \quad , \quad x \in \mathbb{R} \quad , \quad t>0. \tag{4a}$$



Lo que vamos a hacer es cambiar de coordenadas. Llevar a la forma **canonica**.

La nueva ecuacion queda

$$b^*(\xi, \eta) u_{\xi \eta} + d^*(\xi, \eta) u_\xi + e^*(\xi, \eta) u_\eta + f^*(\xi, \eta) u = g^*(\xi, \eta) .$$
La idea es pues eliminlar los terminos con coefficientes $a(\xi, \eta)$ y $c(\xi, \eta)$

La transformacion
$$\xi = \xi(x,t) \quad , \quad \eta = \eta(x, t) \tag{5} $$

Para este cambi de coordenadas sea regular (se pueda hacer sin infinitos)
el jacobiano debe ser distinto de cero.

$$ J = \begin{vmatrix} \xi_t & \xi_x \\ \eta_t & \eta_x \end{vmatrix} \ne 0  .$$


Comencemos la busqueda de todos los coeficientes estrella, al insertar (5) en
 .


usando la regla de la cadena para derivadas parciales
$$u_t = u_\xi \xi_t + u_\eta \eta_t $$
$$u_x = u_\xi \xi_x + u_\eta \eta_x \tag{6} $$

Ahora aplicando la regla de la cadena de nuevo.

\begin{eqnarray}
u_{tx} &=& (u_t)_x  = (u_\xi \xi_t + u_\eta \eta_t)_x \\
&=& (u_\xi \xi_t )_x + (u_\eta \eta_t)_x \\
&=& (u_\xi)_x \xi_t + u_\xi \xi_{tx} + (u_\eta)_x \eta_t + u_\eta \eta_{tx}
\end{eqnarray}
Recordemos ecuacion (6)

$$u_{tx} = (u_{\xi \xi} \xi_x + u_{\eta \xi}\eta_x) \xi_t + u_{\xi} \xi_{tx} + (u_{\eta \xi} \xi_x + u_{\eta \eta} \eta_x) \eta_t + u_{\eta} \eta_{tx} .$$


Escribo la ecuacion anterior con otro orden

$$u_{tx} = u_{\xi \xi} \xi_t \xi_x + u_{\xi \eta} (\xi_t \eta_x + \xi_x \eta_t) + u_{\eta \eta} \eta_t \eta_x + u_\xi \xi_{tx} + u_\eta \eta_{tx} .$$

Lo que sigue es $u_{tt}$ y $u_{xx}$, podemos copiar/pegar/editar, combiando primero $x$ por $t$ y despues $t$ por $x$.

$$u_{tt} = u_{\xi \xi} \xi_t^2 + 2 u_{\xi \eta} \xi_t \eta_x  + u_{\eta \eta} \eta_t \eta_t + u_\xi \xi_{tt} + u_\eta \eta_{tt} .$$


$$u_{xx} = u_{\xi \xi} \xi_x \xi_x + 2 u_{\xi \eta} \xi_x \eta_x  + u_{\eta \eta} \eta_x \eta_x + u_\xi \xi_{xx} + u_\eta \eta_{xx} .$$

Insertamos todo esto en  (4a) que es
$$a(x,t) u_{tt} + 2 b(x,t) u_{tx} + c(x,t) _{xx} + d(x,t) u_t + e(x,t) u_x + f(x,t) u = g(x,t)  \quad , \quad x \in \mathbb{R} \quad , \quad t>0. \tag{4a}$$


y encontramos


$$ a^* u_{\xi \xi} + 2 b^* u_{\xi \eta}  + c^* u_{\eta \eta} + d^* u_{\xi}  + e^* u_\eta  + f^* u = g^*.$$
donde





\begin{eqnarray}
a^* &=& a \xi_t^2 + 2 b \xi_t \xi_x + c \xi_x^2 \\
b^* &=& a \xi_t \eta_t + b ( \xi_t \eta_x + \xi_x \eta_t) + c \xi_x \eta_x \\
c^* &=& a \eta_t^2 + 2 b \eta_t \eta_x + c \eta_x^2 \\
d^* &=& a \xi_{tt} + b  \xi_{tx} + c \xi_{xx} +  d \xi_t + e \xi_x \\
e^* &=& a \eta_{tt} + b \eta_{tx} + c \eta_{xx} + d \eta_t + e \eta_x \\
f^* &=& f \\
g^* &=& g \\
\end{eqnarray}
