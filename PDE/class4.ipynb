En este problema existe una constante. Asumamos que el punto mas alto es $c-1$, $c>1$. Las lineas van de $c-1$ a $1$, de $c-2$ a $2$, de... $c-n$ a $n$.

De forma que


\begin{eqnarray}
 1 + c - 1 &=& c \\
 2 + c - 2 &=& c \\
 &\vdots & \\
 n + c - n &=& c.
\end{eqnarray}

Todas las lineas son de la forma

$$y = \alpha x + b . \tag{1}$$
donde $\alpha$ es la pendiente y $b$ el intercepto. Pensamos que la **familia** esta parametrizada por dos parametros $\alpha, b$, pero no, por que $c$ es constante y podemos elimitar $b$ con esta constante. Aca $c$ es la suma de los intercptos.

Eliminemos $b$ de esta ecuacion.

**Interceptos**:

* $x=0$,  $y=b$
* $x = (y-b)/\alpha$, de forma para $y=0$,
$$ x = -b/\alpha$$.

La suma de los dos interceptos es constante $c$


$$b - \frac{b}{\alpha} = c$$
$$b \left ( 1 - \frac{1}{\alpha} \right ) = c$$


$$ b= \frac{c}{1 - 1/\alpha} = \frac{c \alpha}{\alpha -1 }$$

Volvemos a (1)

$$ y = \alpha x + \frac{c \alpha}{\alpha -1} $$

Recordamos la clase anterior $F(x,y,\alpha) = y - \alpha x - \frac{c \alpha}{\alpha -1 } = 0 $. Recuerden la condicion $\frac{\partial F}{\alpha} = 0$

Tenemos el siguiente sistema


\begin{eqnarray}
 y &=& \alpha x + \frac{c \alpha}{\alpha -1 } \\
 \frac{\partial F}{\partial \alpha} &=& -x - \frac{(\alpha -1)c - \alpha c}{(\alpha -1)^2}
 = -x - \frac{\cancel{\alpha c} - c - \cancel{\alpha c}}{(\alpha -1)^2}
 = -x + \frac{c}{(\alpha -1)^2} = 0
\end{eqnarray}

Todavia no sabemos quien es $\alpha$? Despejamos $\alpha$


$$ (\alpha -1)^2 = \frac{c}{x} $$

$$ \alpha = \pm \sqrt{\frac{c}{x}} + 1 $$

Regresamos a (1) con $b$ y con $\alpha$ encontrados.



* "+"
\begin{eqnarray}
y &=& \left (  1 + \sqrt{ \frac{c}{x}} \right ) x + \frac{c \left ( 1 + \sqrt{\frac{c}{x}} \right )}{\cancel{1} + \sqrt{ \frac{c}{x}} - \cancel{1}} \\
 &=& \left (  1 + \sqrt{ \frac{c}{x}} \right ) x + c \left ( 1 + \sqrt{\frac{c}{x}} \right ) \sqrt{ \frac{x}{c}} \\
 &=& x + \sqrt{cx}  + \sqrt{cx} + c  \\
 &=& x + 2 \sqrt{cx} + c
\end{eqnarray}

* "-"
$$ y = x - 2 \sqrt{cx} + c .$$



```python
# graficamos en Python.

import numpy as np
import matplotlib.pyplot as plt

def envelope(x, c):
    return x + c + 2*np.sqrt(x*c)

def envelope2(x, c):
    return x + c - 2* np.sqrt(x*c)

# plot
plt.xlabel('x')
plt.ylabel('y')
plt.title("Envelope curves")

c=100 # constante
domain = np.arange(0, c, 0.01)
plt.plot(domain, envelope(domain, c), label = '+')
plt.plot(domain, envelope2(domain, c), label = '-')
plt.show
```

```python
plt.figure(figsize=(10,10))

# ciclo para dibujar las rectas
c=100
for i in range(1,c,2):
    x_values=[i,0]
    y_values=[0, 99-i]
    plt.plot(x_values, y_values, 'b')


# titutlos
plt.xlabel('x')
plt.ylabel('y')
plt.title("Envelope curves")

# show grid
plt.grid(True)

# aspect ratio
plt.gca().set_aspect('equal', adjustable='box')

# graficamos el envelope
# plt.plot(domain, envelope(domain, c), label = '+')
plt.plot(domain, envelope2(domain, c), label = '-', color='r')


plt.show()
```

# Algorimo general para las ecuaciones PDE no lineales de primer orden
La ecuacion general esta dada por

$$F(x,y,u, u_x, u_y)=0 .$$
Generalmente $p=u_x$, $q=u_y$.
la ecuacion queda
$$F(x,y,u, p, q)=0 .$$

Notas:

* La geometria del problema no lineal es mas complicada que el quasi-lineal. Vamos a ver dos conceptos geometricos nuevos:
    * Monge cones (conos de Mongie)
    * Strip characteristics (bandas caracteristicas).

* Es probable encontrar problemas de
    * existencia. Por ejemplo $p^2 + q^2 +1 =0$, no tiene solucion
    * unicidad. Si las condiciones iniciales no se definen correctamente puede haber un numero infinito de soluciones.

El problema IVP, es basicamente el mismo. Debemos proporcionar una curva **inicial** en el plano $xy$.  Debemos encontrar la function integral $u=u(x,y)$


## Monge cones:
Cual la idea de "Monge cones". Se asume que la ecuacion
$$F(x,y,u,p,q)=0 \tag{2}$$
tiene fijos $x=x_0, y=y_0, u=u_0$ pero $p,q$ son variables. Asuamos
que $q = q(p)$ ($q$ es funcion de p). Entonces (2) es una **familia**
de ecuaciones para $p$.

* Fijamos el punto $(x_0, y_0, u_0)$ en la superficie integral.
* En ese punto tenemos un plano. El plano tangente a la superficie integral esta dado por


$$p (x - x_0) + q (y - y_0) - (u - u_0) = 0 .$$





Si permitimos que $p,q$ sean variables y $q=q(p)$ esto es una one-paramter-familia de planos, todos pasando por el punto comun $(x_0, y_0, u_0)$ y la familia
depende de $p$. Esa familia tiene un "envelope" que es un cono.

Que sean conos esta por verse, pero podemos intuir que lo son por lo siguiente

* Todos los planos comparten el mismo punto $(x_0, y_0, u_0)$.
* La interseccion de dos planos vecinos es una recta.



image.png

Al momento tenemos dos caracteristicas importantes de los conos de Monge:

1. La familia de planos tangentes
$$ p(x-x_0) + q(y - y_0) = u - u_0 \tag{3} .$$
2.  La ecuacion PDE functional
$$ F(x_0, y_0, u_0, p, q) = 0  \tag{4} .$$
que proporciona una relacion implicita entre $p$ y $q$ que se podria volver
explicita $q=q(p)$.

Derivamos la ecuacion (3) con respecto a $p$

$$ (x - x_0) + \frac{dq}{dp} ( y -y_0) = 0.  \tag{5} $$
Asumimos que $F_p, F_q$ no son cero simultaneamente. Es decir que $F_p^2 + F_q^2 \ne 0$. Sin perdida de generalidad podemos asumir que $F_q \ne 0$.


Diferenciamos (4) con respecto a $p$. (regla de la cadena)

$$ F_p  + F_q \frac{dq}{dp} = 0.$$
de forma que
$$ \frac{dq}{dp} = - \frac{F_p}{F_q}$$
insertamos esto en la ecuacion (5)
$$ (x - x_0) - \frac{F_p}{F_q} ( y -y_0) = 0.  $$
$$ (x - x_0) =  \frac{F_p}{F_q} ( y -y_0)  $$

$$ \frac{x-x_0}{F_p} = \frac{y-y_0}{F_q} .$$

Ahora usamos la siguiente propidad (tarea)
Si

$$\frac{a}{b} = \frac{c}{d} .$$
entonces
$$ \frac{a}{b} = \frac{pa + qc}{pb + qd} .$$
donde $p,q$ son arbitrarios (no cero simultaneamente).
En esta ecuacion reemplazamos

$$ a= x - x_0 \quad, \quad b=F_p \quad , \quad c=y-y_0 \quad , \quad d=F_q .$$
encontramos

$$ \frac{x -x_0}{F_p} = \frac{y - y_0}{F_q} = \frac{p (x-x_0) + q(y -y_0)}{p F_p + q F_q} $$

Ademas como $p(x-x_0) + q(y-y_0) = u-u_0$ (ecuacion (3), entonces

$$ \frac{x -x_0}{F_p} = \frac{y - y_0}{F_q} = \frac{u - u_0}{p F_p + q F_q}  \tag{6} $$

Esta ecuacion representa una familia de lineas parametrizadas por $p$ todas ellas pasando por el punto $(x_0, y_0, u_0)$. Esto es un cono.

[Este enlace](http://www.mndcollegerajur.org/uploads/department/cone.pdf) nos muestra como construir el cono a partir de la curva generadora (6).

En resumen: El cono se halla eliminando $p,q$ de las ecuaciones

\begin{eqnarray}
F(x_0, y_0, u_0, p, q) &=& 0 \\
u - u_0 = p (x - x_0) + q (y - y_0) \tag{6} \\
\frac{x -x_0}{F_p} = \frac{y - y_0}{F_q} \tag{7}
\end{eqnarray}

**Ejemplo 1:**
Encuentre el cono Monge de la ecuacion **eikonal** $p^2 + q^2 = 1$,
en el punto $(x_0,y_0, u_0)=(0,0,0)$.

La ecuacion funcional es $F(x_0, y_0, u_0, p, q)= p^2 + q^2 - 1 \equiv 0$.

Despejamos $q$ en terminos de $p$

$$ q = \pm \sqrt{1 - p^2} .$$
De la ecuacion del plano tangente (6) en el punto

$$ u = px + qy \tag{8} .$$
y de la equation (7), tenemos que
recuerden que $F_p = 2 p$, $F_q=2q$.
$$ \frac{x}{2 p} = \frac{y}{2 q} .$$
Necesitamos eliminar $p,q$.

$$ \frac{x}{2 p} = \frac{y}{2q} = k .$$
entonces

$$ p = \frac{x}{2 k} \quad , \quad q = \frac{y}{2k} .$$
Ahora sustituimos esto en la (8)

$$ u = \frac{x^2}{2k} + \frac{y^2}{2 k} = \frac{x^2 + y^2}{2k} \tag{9}.$$

Todavia no sabemos quien es $k$?

Para eso usamos que $p^2 + q^1 = 1$.

$$ \frac{x^2}{4 k^2} + \frac{y^2}{4 k^2} = 1 .$$
de donde

$$ x^2 + y^2 = 4 k^2.$$
y
como

$$ x^2 + y^2 = 2 k u \quad , \quad \text{de la ecuacion (9)} $$,
entonces
$$  4 k^2 = 2 k u.$$
$u = 2k$

$$ x^2 + y^2 = 2 ku = u^2$$

$$ x^2 + y^2 = u^2$$




```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# definamos el cono en coordenadas polares.
theta = np.linspace(0, 2 * np.pi, 100)
r = np.linspace(0, 2, 100)
# malla
theta, r = np.meshgrid(theta, r)
# coordenadas cartesianas
x = r * np.cos(theta)
y = r * np.sin(theta)
z = r

# el plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# etiquetas
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

Que pasa con los conos Monge cuando la PDE es quasi-linear.

Recuerden que las ecuaciones quasi-lineales se representan como

$$ a(x,y,u) p + b(x,y,u) q = c(x,y,u).$$
Aca $F(x,y,u,p,q)=a(x,y,u)p + b(x,y,u)q - c(x,y,u) \equiv 0$.
El cono debe contener el vector tangente a lo largo de las caracteristicas
$\mathbf{v} = [ a(x,y,u), b(x,y,u), c(x,y,u)]$. Al comparar vemos que


$$F_p = a(x,y,u) \quad , \quad F_q = b(x,y,u) , \quad , \quad u = p F_p + q F_q .$$

Ahora bien, $p F_p + q F_q = a(x,y,u) p + b(x,y,u) q = c(x,y,u) $
de forma que tenemos que el cono es generado por la linea


$$\frac{x-x_0}{a(x_0, y_0, u_0)} = \frac{y - y_0}{b(x_0, y_0, u_0)} = \frac{u-u_0}{c(x_0, y_0, z_0)}.$$


Si escribimos diferenciales en la ecuacion (6)

$$\frac{dx}{F_p} = \frac{dy}{F_q} = \frac{du}{p F_p + q F_q} .$$
y estas son las caracteristicas de la superficie integral.

Si usamos la notacion de las caracteristicas, parametricas con parametro $\sigma$,


\begin{eqnarray}
\frac{d x(\sigma)}{d \sigma } = F_p   \quad , \quad
\frac{d y(\sigma)}{d \sigma } = F_q   \quad , \quad
\frac{d u(\sigma)}{d \sigma } = p F_p    + q F_q \tag{10}
\end{eqnarray}

Estas equaciones ordinarias coinciden con aqueyas en quasi-linear PDE si

$$a(x,y,u)= F_p \quad , \quad b(x,y,u)= F_q \quad , \quad c(x,y,u)= pF_p + q F_q .$$

El problema es que $F_p, F_q$ pueden depender de $p,q$ y no podemos usar la teoria de las clases anterior.

**Ejemplo 2**: asuma la PDE no lineal $u_x^2 + u_y^2  = u$. Resuelva esta ecuacion.

$$F(x,y,u,p,q) = p^2 + q^2 - u \equiv 0 .$$
Necesitamos

$$ F_p = 2 p \quad, \quad F_q=2 q, entonces.$$
De (10)
\begin{eqnarray}
\frac{d x(\sigma)}{d \sigma } = 2p   \quad , \quad
\frac{d y(\sigma)}{d \sigma } = 2 q   \quad , \quad
\frac{d u(\sigma)}{d \sigma } = 2 p^2    + 2 q^2
\end{eqnarray}
pero $p^2 + q^2 = u$,
\begin{eqnarray}
\frac{d x(\sigma)}{d \sigma } = 2p   \quad , \quad
\frac{d y(\sigma)}{d \sigma } = 2 q   \quad , \quad
\frac{d u(\sigma)}{d \sigma } = 2 u
\end{eqnarray}

Pero no tenemos $p,q$, como las obtenemos? lo vemos en esta media hora.


Asuma que $p=p(x(\sigma), y (\sigma))$
Usamos la regla de la cadena en $p= u_x=u_x(x(\sigma), y(\sigma))$

$$p_\sigma =  u_{xx} x_{\sigma} + u_{xy} y_{\sigma} = u_{xx} F_p +u_{xy} F_q .$$
donde usamos la (10).
$$q_\sigma =  u_{yx} x_{\sigma} + u_{yy} y_{\sigma} = u_{yx} F_p +u_{yy} F_q .$$

En terminos de $p$, $q$, $u_{xx} = p_x, u_{yy}=q_y$, $u_{xy}= p_y$

$$ p_\sigma = p_x F_p + p_y Fq  \tag{10a}$$
$$ q_\sigma = q_x F_p + q_y Fq \tag{10b}$$


Necesitmoas eliminar $p_x, p_y, q_x=p_y$,
Saquemos la derivada de $F=F(x,y,u,p,q)$ con respecto a $x,y$


$$F_x + p F_u + p_x F_p + q_x F_q = 0 . \tag{11}$$
$$F_y + q F_u + p_y F_p + q_y F_q = 0 .$$
Pero ahora las cruzadas son iguales $p_y=q_x$

$$F_y + q F_u + q_x F_p + q_y F_q = 0 \tag{12}.$$
reemplazando (10a) y (10b) en la (11) y la (12).

$$ \frac{dp}{d \sigma} = - F_x - p F_u .$$
$$ \frac{dq}{d \sigma} = - F_y - q F_u .$$

Completamos el siguiente sistema de ODEs

\begin{eqnarray}
\begin{cases}
\frac{dx}{d \sigma} = F_p \\
\frac{dy}{d \sigma} = F_q \\
\frac{du}{d \sigma} = p F_p + q F_q \\
 \frac{dp}{d \sigma} = - F_x - p F_u  \\
 \frac{dq}{d \sigma} = - F_y - q F_u
\end{cases}
\end{eqnarray}
