## Notas sobre el teorema de la **funcion inversa**.

*Ejemplo 1**: (una dimension). Considere la funcion $f(x)=x^2/2$, $y=x^2/2$. Tiene inversa? $x=f^{-1}(y)$. La respuesta es, depende

$$(-2)^2/2=(+2)^2=2 .$$
para este caso hay dos preimagenes $\pm2$. Entonces el problema donde esta?
Esta en $x=0$ y vamos a ver por que.

* Si $f'(x) > 0$: la funcion es monotona creciente y podemos definir un intervalo al redededor del cual la funcio tiene inversa.
* Si $f'(x) < 0$: La funcio es mononota decreciente y tambien podemos definir un intervalo al redededor del cual la funcion tiene inversa (es uno-a-uno y sobreyectiva).


* Si $f'(x)=0$, la funcio es decreciente a la izquierda y creciente a la derecha y es **imposible** encontrar un intervalo donde la funcion tenga inversa (no es uno-a-uno).

De hecho
$x=\sqrt{2 y}$

$$f'(y)=  \frac{dx}{dy} = \frac{1}{2 y} .$$

Se presentan los siguientes problemas: $x=0$.

* La funcion no es uno-a-uno (por que es creciente a la derecha y decreciente a la izquierda
* La derivada de la inversa no existe.  La derivada $dy/dx=0$, en $x=0$,
pero la derivada de la inversa es $dx/dy=1/\sqrt{2 y}$.
* Mientras que la pendiente de la funcion directa es 0, la pendiente de la funcion inversa es $dx/dy = 1/(dy/dx) = 1/0$ es vertical.

**Ejemplo 2** (2D).

Asumamos el siguiente sistema:

$f: \mathbb{R}^2 \to \mathbb{R}^2$.

\begin{eqnarray}
\begin{cases}
a x + by = u \\
cx + dy = v \tag{1}
\end{cases}
\end{eqnarray}

Podemos definir functions

$$ f_1(x,y) = ax + by - u \equiv 0 $$
$$ f_2(x,y) = cx + dy - v \equiv 0 $$

La derivada de este campo vectorial esta dada por la matriz Jacobiana

$$ J_M = \begin{pmatrix} \frac{\partial f_1}{\partial x} & \frac{\partial f_1}{\partial y}  \\ \frac{\partial f_2}{\partial x} & \frac{\partial f_2}{\partial y} \end{pmatrix}  .$$

El sistema (1) se puede escribir como

$$A \bmx = \bmb .$$
donde

$$ A = \begin{pmatrix} a & b \\ c& d \end{pmatrix} .$$

O sea que para que el sistema tenga **inversa** se necesita que
$\det A \ne 0$, pero $A=J_M$ o sea que debemos tener que

$$ \det J_M = \begin{vmatrix} \frac{\partial f_1}{\partial x} & \frac{\partial f_1}{\partial y} \\ \frac{\partial f_2}{\partial x} & \frac{\partial f_2}{\partial y} \end{vmatrix} = J \ne 0 .$$

En este caso $x = A^{-1} \bmb$.

El caso no lineal

$$f(x,y) = \binom{x^2}{xy} .$$
Dejo como ejercicio que encuentren los puntos para los cuales $J=0$. Y que pasa en estos puntos?


**Ejemplo 3**: El caso de $n$ dimensiones.
Piense en el problema $A \bmx = \bmy$, con $A \in \mathbb{R}^{n \times n}$,
$\bmx, \bmy \in \mathbb{R}^n$.


En forma granulada (en indices)


$$ \sum_{j=1}^n a_{ij} x_j = y_i \quad x_i, y_i \in \mathbb{R}, i =1,2, \cdots, n .$$


DEfinimos las funciones auxiliares


$$f_i(x_1, x_2, \cdots, x_n) = \sum_{j=1}^n a_{ij} x_j - y_i \equiv 0 .$$

Ahora encontramos que


$$ a_{ij} = \frac{\partial f_i}{\partial x_j} .$$

De nuevo la matriz Jacobiana es $A$, es decir $J_M = A$, y
necesitamos qeu $\det J_M =\det A = J \ne 0$.
En este caso existe la inversa de $x$ en funcion de $y$ y es

$$ \bmx = A^{-1} \bmy $$

El **teorema de la funcion inversa** dice que para exista la inversa de $f(\bmx)=\bmy$ con $f=(f_1, f_2, \cdots, f_n)$ necesitamos que

$$ J = \det \left (  \frac{\partial f_i}{\partial x_j } \right )  \ne 0 .$$

## Descripcion general del algortimo de las **caracteristicas** para resolver PDEs de primer orden.

La ecuacion lineal PDE de primer orden esta dada por

$$a(x,y) u_x + b(x,y) u_y + c(x,y) u = f(x,y) .$$

Podemos redefinir

$$c(x,y,u) = -c(x,y) u + f(x,y) .$$
para escribir
$$a(x,y) u_x + b(x,y) u_y = c(x,y,u)  \tag{2} .$$

Definamos $F=u(x,y)-u \equiv 0$. De calculo multidimensional

$$\nabla F = [u_x, u_y, -1] .$$
Rercuerde que $\bmn=\nabla F$ es normal a la superficie $u=u(x,y)$ y podemos escribir el sistema como

$$[a(x,y), b(x,y), c(x,y)] \cdot [u_x, u_y - 1] = 0 .$$
De forma que el vector

$$\bmv = [a(x,y), b(x,y), c(x,y)] .$$ es ortogonal al gradiente $\nabla F$.
o sea que es tangente a la superficie $u=u(x,y)$.

La superficie solucion (integral surface) se escribir de forma parametrica basada en ls parametros $\tau, \sigma$ como


$$ x = x(\sigma, \tau) \quad , \quad   y=y (\sigma, \tau) \quad , \quad u=u(\sigma, \tau) .$$

La superficie la escribo usando el radio vector $\bmr$

$$\bmr(\sigma, \tau) = [x(\sigma, \tau), y(\sigma, \tau), u(\sigma, \tau)] .$$

Vamos a escoger la tangnte a esta curva en la direccion del vector $\bmv$.

La tangente a la curva , a lo largo de $\sigma$ (fijando $\tau$) esta dada por

$$\bmt = \frac{d \bmr}{d \sigma} = \left [ \frac{dx}{d \sigma}, \frac{dy}{d \sigma}, \frac{du}{d\sigma}
  \right ] .$$

Si queremos que este vector este a lo largo de la curva $\bmr(\sigma)$ en
la direccion del vector $\bmv$ entonces ambos son paralelos

O sea que

$$\bmt = \lambda \bmv .$$

image.png

De aca sale que

$$ \frac{dx}{d \sigma} = \lambda a (x,y) \quad , \quad \frac{dy}{d \sigma} = \lambda b(x,y) \quad , \quad \frac{du}{d\sigma} = \lambda c(x,y,u) .$$

Varios puntos

* Las dos primeras ecuaciones mestran que las curvas $x(\sigma), y(\sigma)$ son planas. Si los coeficientes son constantes esas curvas son lineas rectas.
Como ejercicio les pregunto que pasa si $a(x,y)=y$, $b(x,y)=x$, las
caracteristicas no son rectas.


Como se resuelve el sistema.

*  Se resuelveln las dos primeras para obtener $x,y$.
*  Se se resuelve la ultima, conocidos $x,y$.
*  Como $x(\sigma, \tau), y(\sigma, \tau)$, la solucion $u=u(\sigma, \tau)$,
Para que quede $u(x,y)$ toca inverti estas funciones y necesitamos aplicar el teorema de la funcion inversa.

De las primeras dos ecuaciones y las ultimas

$$ \frac{dy}{dx} = \frac{b(x,y)}{a(x,y)} \quad , \quad  \frac{du}{dx} = \frac{c(x,y,u)}{a(x,y)} $$

Como escojemos $\lambda$?

Un caso interesante, si

$$ \lambda =  \frac{1}{\sqrt{a^2 + b^2 + c^2}} .$$
entonces

$$ \| \bmt \| = \sqrt{ \frac{dx}{d \sigma}^2 +  \frac{dy}{d \sigma}^2 + \frac{du}{d \sigma}^2} = \frac{\sqrt{a^2 + b^2 + c^2}}{\sqrt{a^2 + b^2 + c^2}} = 1  $$

Es decir que esta escogencia de $\lambda$ me dice que el parametro $\sigma$ es **longitud de arco**.


Nosotros escogemos $\lambda=1$.


## Especificaion de datos.
Recuerden las curvas $\sigma, \tau$.

El parametro inicial es $\sigma=0$. Necesitamos proveer las curvas (dato iniciales)

$$ x = x( 0, \tau) \quad, \quad y=y(0, \tau) \quad , u=u(0, \tau) .$$

Para cada $\tau$ fijo, resolvemos $x=x(\sigma, \tau)$, $y=y(\sigma,\tau)$
y $u=u(\sigma, \tau)$, Pero necesitamos $u(x,y)$.


Como necesitamos $u(x,y)$ debemos invertir el sistema

\begin{eqnarray}
x = x(\sigma, \tau) \\
y = y(\sigma, \tau) \\
\end{eqnarray}

Para esto se requiere

$$ J = \begin{vmatrix} x_{\sigma} & x_{\tau} \\ y_{\sigma} & y_{\tau} \end{vmatrix} \ne 0  .$$

En las siguientes lineas vamos a ver que $J=0$ crea problemas en la inversion, y que signifca que $J=0$.

Supongamos que $J=0$, entonces


$$ x_\sigma  y_\tau  - y_\sigma x_\tau = 0 \quad , \quad \sigma=0. \tag{4}$$

Podemos escribir esto como

$$ \frac{x_\tau}{y_\tau} = \frac{x_\sigma}{y_\sigma} = \frac{a}{b}  \quad , \quad \sigma=0.$$

Pensamos $x=x(\tau)$, $y=y(\tau)$ por que $\sigma=0$ es constante con respecto a $\sigma$

Tomemos la primera y la tercera de (4)

$$ \frac{dx}{d \tau} b - \frac{dy}{d \tau} a = 0  \quad \sigma=0.$$


El vector tangente a esta curva localmente esta dado por
$\bmt = (dx/d \tau, dy/d \tau)$ (curva plana). Queremos (recuerde la curva azul en el tablero) que este vector no este en las direcciones de $[a,b]$. Matematicamente que no sean paralelas quiere decir que los dos vectores en $\mathbb{R}^2$ son linealmente independientes. Es decir el determinante no puede ser cero.


$$ \begin{vmatrix}  \frac{dx}{d \tau} & a \\ \frac{dy}{d \tau} & v\end{vmatrix}
 \frac{dx}{d \tau} b - \frac{dy}{d \tau} a \ne 0.$$

Retornamos a $J=0$.

$$ \frac{x_\tau}{a} = \frac{y_\tau}{b} .$$

Recuerden $u=u(x(\tau), y(\tau))$. Usando la regla de la cadena

$$u_\tau = u_x x_\tau + u_y y_\tau \quad \sigma=0  .$$

Tenemos entonces

\begin{eqnarray}
u_{\tau} &=& y_\tau \left [ \frac{x_\tau}{y_\tau} u_x + u_y     \right ] \\
&=& y_\tau \left [ \frac{a}{b} u_x + u_y   \right ] \\
&=& \frac{y_\tau }{b} \left [ a u_x + b u_y   \right ] \\
&=& \frac{y_\tau}{b} c
\end{eqnarray}

Pasando a dividir en la ultima expresion

$$ \frac{x_\tau}{a} = \frac{y_\tau}{b} = \frac{u_\tau}{c} .$$
