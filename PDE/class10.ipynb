# Continuacion en SLO (Sturm-Liouville Operators)
En la clase anterior mostramos que si $B=\{ \varphi_i(x) \}_{i=1}^{\infty}$ es una base (esto lo explicamos mejor abajo) en el espacio de funciones que nos atanhe, entonces

Cualquier funcion en el espacio $v$ se puede escribir como la combinacion lineal
$$ v= \sum_{i=1}^{\infty} c_i  \varphi_i $$
(esta es la **serie de Fourier**) y que los coeficientes $c_i$ se pueden
extraer de la proyeccion de $v$ a las funciones base. Es decir,

$$ c_i = \langle v, \varphi_i \rangle $$

La pregunta es si son suficientes todas las auto-funciones $\varphi$? Definimos
entonces los conjuntos **completos**.


**Definicion**: Un espacio ortonormal $B=\{ \varphi_n \}_{n=1}^\infty $ es **completo** si para todo vector (funcion) $v$ esta se puede expresar como

$$v = \sum_{n=1}^\infty \langle v, \phi_i \rangle \phi_i \tag{1} $$

La ecuacion (1) se llama **series de Fourier** y el coeficiente
$\langle v, \phi_i \rangle$ se llama **coeficiente de Fourier**.

**Ejemplo**: considere el espacio de secuencias $\ell_2$ donde los vetores son de la forma $$v=(a_1, a_2, \cdots, a_n, \cdots ) $$, tales que

$$\sum_{i=1}^\infty | a_i |^2 < \infty  $$
La base
$$B = \| \phi_i | (\phi_i)_j = \delta_{ij} \} $$
es **completo** en $\ell_2$. Cada elemento $\phi_i$ de la base tiene un 1 en la
posicion $i$ y 0 en los demas lugares. El conjunto es completo por que cualquier $v \in \ell_2$ se puede escribir como

Aca

\begin{eqnarray}
\delta_{ij} = \left \{
    \begin{array}{cc}
    1 & \text{si } i = j \\
    0 & \text{si } i \ne 0
    \end{array}
    \right .
\end{eqnarray}
Es la representacion indizal de la matriz identidad.


$$v = \sum_{i=1}^\infty \langle v, \phi_i \rangle \phi_i $$
como vimos en el tablero a mano alzada.

Ahora quitemos de $B$ el elemento $\phi_1$, entonces el vector
$(1, 0, 0, \cdots )$ no se puede escribir como combinacion lineal
de los vectores $B- \phi_1$.

La ecuacion (1) es muy importante para calcular aproximaciones

$$ v \approx \sum_{i=1}^N \langle f, \varphi_i \rangle \phi_i $$

Aun mas:

$$\| v \|^2 = \langle v, v \rangle = \sum_{i=1}^\infty \sum_{j=1}^\infty \langle f, \phi_i \rangle \langle f, \phi_j \rangle \langle \phi_i, \phi_j \rangle  =  \sum_{i=1}^\infty \sum_{j=1}^\infty \langle f, \phi_i \rangle \langle f, \phi_j \rangle \delta_{ij} = \sum_{i=1}^\infty \langle f, \phi_i \rangle^2  $$
Este seria el teorema de **Pitagoras** generalizado. Si truncamos a algun $N$
tenemos

$$\| v \|^2 \le  \sum_{i=1}^N \langle f, \phi_i \rangle^2 $$

# Metodo de separacion de variables para la ecuacion de onda 1D.
Voy a poner una lista d elo que se hace

* Asuma que $u(x,t)=F(x)G(t)$ (un anzats). Y por que funciona? (ecuaciones lineales)

* Insertar esta ecuacion en la PDE
* Encontrar la solucion $F(x)$ usando condiciones de bordo
* Usar el principio de superposicion. Es decir sumar todas las soluciones
$L(\sum_i \varphi_i)=\sum_i L (\varphi_i)$

* Usar la condicion inicial y al proyectarla  a las bases (autofunciones) obtener los coeficientes). Ver primeros minutos de esta clase.

veamos un ejemplo.
Recuerden la ecuacion de onda en $[0, \ell]$

image.png

Si quisieramos resolver la ecuacion de onda para $t \gg 1$ o hasta $t \in \infty$, necesitariamos extender la condicionon inicial a periodica entre $(-\infty, \infty)$, con periodos de la forma $((-n-1) \ell, -n \ell)$ para los negativos y $(n \ell , n+1 \ell)$ y esto es muy dispendioso.


El metodo de separacion de variables viene al rescate.

El problema es:

\begin{eqnarray}
\begin{cases}
u_{tt} - c^2 u_{xx} = 0 \\
u(0,t) = u(\ell , 0 ) = 0 \\
u(x,0) = \varphi(x) \quad, u_t(x,0) = \psi(x) = 0.
\end{cases}
\end{eqnarray}

Resolvamos esto por separacion de variables

* Primer paso, asumir que la solucion se escribe con separacion de varaibles

$$ u(x,t) = F(x) G(t)  $$

* Insertar estas en la ecuacion de onda

$$ u_{tt} = F(x) G''(t) \quad , \quad u_{xx} = F''(x) G(t) $$
De la ecuacion de onda encontramos

$$F(x) G''(t) = c^2 F''(x) G(t) $$

Reordenando

$$\frac{F''(x)}{F(x)} = \frac{1}{c^2} \frac{G''(t)}{G(t)} = k $$
Una de las consecuencias grandes de separacion de variables es que $k$ es constante (no puede ser solo funcion de $x$ y solo funcion de $t$).
De aca salen dos ODEs.

$$F''(x) = k F(x)  \quad , \quad G''(t) = k c^2 G(t)  $$

Vamos a considerar 3 casos $k<0, k=0, k>0$
    
1. $k=0$:
Las ecuaciones se convierten en
$$F''(x) = 0   \quad , \quad G''(t) = 0  $$

De forma que $F''(x)=0$ es decir

$$F(x) = C_1 x + C_2 $$
$$G(x) = C_3 x + C_4 $$

Aplicamos las condiciones de frontera $u(0,t)=u(\ell, t) = 0$

$$F(0) = C_1 0 + C_2 = C_2 = 0 $$
Pero entones $F(\ell)=0$, $C_1 \ell + C_2 = C_1 \ell = 0$. $C_1 = 0$
Es decir $F(x)=0$. Piense en que es una linea y tiene dos ceros (en $y$) en dos puntos distintos. Esta solucion $u(x,t)=0$ es la trivial  y no nos interesa.

2. $k>0$, llamemos, por conveniencia $k=\lambda^2$, encontramos las siguientes dos ecuaciones

$$F'' - \lambda^2 F= 0  $$
$$G'' - \lambda^2 c^2 G= 0  $$
Llamemos $D=d/dx$ y tenemos para la primera ecuacion\

$$(D^2 - \lambda^2) F = 0  $$
Factorizando

$$(D - \lambda)(D+ \lambda) F = 0  $$
Salen dos ecuaciones

$$ \frac{dF}{dx} = \lambda F \implies F= C_1 \mathrm{e}^{\lambda x}$$
$$ \frac{dF}{dx} = -\lambda F \implies F= C_2 \mathrm{e}^{-\lambda x}$$
Combinamos las dos (superposicion)
Tenemos

$$f(x) = C_1 \mathrm{e}^{\lambda x} + C_2 \mathrm{e}^{-\lambda x} $$

* Aplicamos las condiciones de frontera.
$u(0,t)=u(\ell,t)=0$, pero $u=FG$, asumamos que $G \ne 0$, por que
entonces $u = 0$ y esa no es gracia.
\begin{eqnarray}
F(0) &=& C_1 + C_2 = 0 \\
F(\ell) &=& C_1 \mathrm{e}^{\lambda \ell} + C_2 \mathrm{e}^{-\lambda \ell} = 0
\end{eqnarray}

De la primera ecuacion $C_2=-C_1$ y sustituyendo esto en la segunda


$$ C_1( \mathrm{e}^{\lambda \ell} - \mathrm{e}^{-\lambda \ell} ) = 0$$
Tarea: prueben que $ \mathrm{e}^{\lambda \ell} - \mathrm{e}^{-\lambda \ell} \ne 0$ de forma $C_1=0$, $C_1=-C_2=0$ entonces $F=0$ entonces $u(x,t)=0$ que es la trivial.








Hasta el momento vimos que $k=0, k<0$ producen soluciones $u(x,t)=0$.
3.  $k<0$, por conveniencia llamemos $k=-\lambda^2$
Las ecuaciones resultantes son

$$F'' + \lambda^2 F = 0  \quad ,  \quad G' + \lambda^2 c^2 G = 0 $$
Resolvemos la primera, llaemos $D=d/dx$, de forma que tenemos
$(D^2 + \lambda^2)=(D + \lambda \mathrm{i})(D - \lambda \mathrm{i})=0 $


\begin{eqnarray}
(D - \mathrm{i} \lambda)F = 0 \implies \frac{dF}{dx} = \mathrm{i} \lambda F \implies F= C_1 \mathrm{e}^{\mathrm{i} \lambda x} \\
(D + \mathrm{i} \lambda)F = 0 \implies \frac{dF}{dx} = -\mathrm{i} \lambda F \implies F= C_2 \mathrm{e}^{-\mathrm{i} \lambda x} \\
\end{eqnarray}
Combinando (superposicion)


$$F = C_1 \mathrm{e}^{\mathrm{i} \lambda x} + C_2 \mathrm{e}^{-\mathrm{i} \lambda x} $$

Aplicamos las condiciones iniciales $u(0,t)=u(\ell, t)=F(0)G(0)=F(\ell) G(\ell)$ no quiero $G$ sea 0$, entonces

$$F(0) = C_1 + C_2 = 0 $$
$$F(\ell) = C_1 \mathrm{e}^{\mathrm{i} \lambda \ell} + C_2 \mathrm{e}^{-\mathrm{i} \lambda \ell} = 0 $$
De la primera $C_2=-C_1$ e insertanto en la segunda encontramos

$$2 C_1 \mathrm{i} \sin \lambda \ell = 0 $$
$C_1 \ne 0$ (para evitar la solucion trivial) entondes

$$\lambda \ell = n \pi \implies \lambda = \lambda_n  = \frac{n \pi}{\ell} $$
Tenemos el siguiente par de autovalores/autofunctiones


$$\lambda_n = \frac{n \pi}{\ell} \quad , \quad F_n(x)= 2 C_1 \mathrm{i} \sin \frac{ n \pi x}{\ell} $$

Resolvamos ahora la ecuacion en tiempo.
De la misma forma

Resolvemos la primera, llaemos $D=d/dt$, de forma que tenemos
$(D^2 + c^2 \lambda^2)=(D + c \lambda \mathrm{i})(D - c \lambda \mathrm{i})=0 $


\begin{eqnarray}
(D - \mathrm{i} c \lambda)G = 0 \implies \frac{dG}{dt} = \mathrm{i} \lambda F \implies G= C_1 \mathrm{e}^{\mathrm{i} c \lambda t} \\
(D + \mathrm{i} \lambda c)G = 0 \implies \frac{F}{dt} = -\mathrm{i} \lambda G \implies G= C_2 \mathrm{e}^{-\mathrm{i} c \lambda t} \\
\end{eqnarray}
Combinando (superposicion)


$$G = C_3 \mathrm{e}^{\mathrm{i} c \lambda t} + C_2 \mathrm{e}^{-\mathrm{i} c \lambda t} $$

Aplicamos las condiciones iniciales $u(0,t)=u(\ell, t)=G(0)G(0)=F(0) G(0)$ no quiero $G$ sea 0$, entonces


$$G(0) = C_3 + C_4 = \varphi(x) $$
De la derivad $u_t(x,0)=0$, $u_t= F(x)G_t(x)$

$$ G_t(0)= \mathrm{i} c \lambda C_3 - \mathrm{i} c \lambda C_4 = 0 $$
$$ G_t(0)= (\mathrm{i} c \lambda)( C_3 - C_4) = 0 $$
Entonces $C_3=C_4$ y la solucion $G(t)$ es

$$G(t) = 2 C_3 \cos \lambda ct $$
Habiamos allado que $\lambda = \lambda_n = n  \pi/\ell$ de froma que

$$G_n(t) = 2 C_3 \cos \frac{n \pi}{\ell} ct $$


Recuerden que $u(x,t)=F(x)G(t)$ entonces

$$ u_n(x,t)= 4 \mathrm{i} C_1 C_3 \sin \frac{n \pi x}{\ell} \cos \frac{n \pi c t}{\ell} $$
aca $u_n$ no es la derivada de $u$ con respecto a $n$, $u_n$ es la $n$-esima solucion



cada autonfuncion $F_n, G_n, u_n$ puede tener coeficientes distintos.  De forma que es mas general pensar que $a_n = 4 \mathrm{i} C_1 C_3$ para la autofuncion de indice $n$.  Y escribimos de forma compacta


$$u_n(x,t) = F_n(x) G_n(t) = a_n \sin \frac{n \pi x }{\ell} \cos \frac{n \pi t}{\ell} $$


La solucion al final es la superposicion de todas estas funciones. Es decir

$$u(x,t) = \sum_{n=1}^{\infty} a_n \sin \frac{\mathrm{i} \pi x}{\ell} \cos \frac{\mathrm{i} \pi t}{\ell}.  \tag{2} $$
Como esto resulto de problemas de SLO autoadjuntos, entonces las funciones
$\sin (n \pi)/\ell $ son ortogonales (automaticamente) sin embargo vamos a probar esto y a encontrar las normas.


Es decir, calculamos, si $f(x) = \sin (n \pi x)/\ell $


$$ \langle f_n, f_m \rangle = \int_0^{\ell} \sin \frac{n \pi x}{\ell} \sin \frac{m \pi x}{\ell} dx $$

Para evaluar esa integral usamos la identidad

$$\sin A \sin B = \frac{\cos(A-B)- \cos (A + B)}{2} $$

Veamos entonces $n \ne m$


\begin{eqnarray}
\langle f_n, f_m \rangle &=& \int_0^\ell \frac{\cos [(n-m) \frac{\pi x}{\ell}] -[\cos( [(n+m) \frac{\pi x}{\ell}]}{2} dx \\
&=&
\frac12 \frac{\ell}{\pi ( n - m) } \left [ \sin(n-m) \frac{\pi x}{\ell}  \right ]_0^\ell +
\frac12 \frac{\ell}{\pi ( n + m) } \left [ \sin(n + m) \frac{\pi x}{\ell}  \right ]_0^\ell = 0
\end{eqnarray}
pues para $x=\ell$ el argumento es $(n \pm m ) \pi$ donde $\sin$ es 0, y lo mismo para $\sin 0 = 0$.

Necesitamos la magnitud para calcular $a_n$

\begin{eqnarray}
\| f_n \|^2 &=& \langle f_n , f_n \rangle \\
&=& \int_0^{\ell} \frac{1 - \cos \frac{2 n \pi x}{\ell}}{2} \\
&=&  \left . \frac12 \right |_0^{\ell} - \left . \frac12 \frac{\ell}{2 n \pi} \sin \frac{2 n \pi x}{\ell} \right |_0^\ell \\
&=& \frac{\ell}{2}.
\end{eqnarray}
Encontramos entonces que

\begin{eqnarray} \langle f_i, f_j \rangle = \frac{\ell}{2} \delta_{ij}
=\left \{
    \begin{array}{cc}
    0 & \text{si } i \ne j \\
    \ell/2 & \text{si } i = j
    \end{array}
    \right .
\end{eqnarray}

* Condicion inicial:
De la ecuacion (3) para $t=0$ tenemos


$$\varphi(x) = \sum_{i=1}^{\infty} a_i \sin \frac{i \pi x}{\ell} $$
Como usamos esto para encontrar los $a_n$? Projectamos la condicion
inicial $\varphi$ sobre las autofunciones.
Es decir

$$\left \langle \varphi(x) ,  \sin \frac{j \pi x}{\ell} \right \rangle = \sum_{i=1}^\infty a_i \left \langle \sin \frac{i \pi x}{\ell},  \sin \frac{j \pi x}{\ell} \right \rangle = \sum_{i=1}^{\infty} a_i \frac{\ell}{2} \delta_{ij} =    a_j \frac{\ell}{2}$$

De forma que

$$a_j = \frac{\ell}{2} \left \langle \varphi(x) , \sin \frac{j \pi x}{\ell}  \right \rangle $$

$$a_i = \frac{\ell}{2} \left \langle \varphi(x) , \sin \frac{i \pi x}{\ell}  \right \rangle = \frac{\ell}{2} \int_0^{\ell} \varphi(x) \sin \frac{i \pi x}{\ell} dx$$


Encontramos que la solucion es:

\begin{eqnarray}
u(x,t) &=& \sum_{i=1}^{\infty} a_j \sin \frac{ i \pi x}{\ell} \cos \frac{i \pi c t}{\ell} \\
a_n &=&  \frac{\ell}{2} \int_0^{\ell} \varphi(x) \sin \frac{i \pi x}{\ell} dx
\end{eqnarray}

Vamos a ver un ejemplo en `Python`.

La condicion inicial

\begin{eqnarray}
u(x,0) = \varphi(x) = \left \{
    \begin{array}{cc}
    0  & \text{for } x < 0 \\
    x  & \text{for } 0 \le x < 0.5 \\
    1 - x  & \text{for } 0.5 \le x < 1 \\
    \end{array}
    \right .
\end{eqnarray}


```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
l=1.0
c=1.0
nx=500
nt=200
x = l*np.linspace(0,1,nx)
t = c*np.linspace(0,1,nt)

def phi(x):
    return np.piecewise(x,[ x<0, (x >=0) & (x<=0.5) , (x >= 0.5) & (x<=1), x>1],
          [lambda x: 0,
           lambda x: x,
           lambda x: -x + 1,
           lambda x: 0
           ])

y=phi(x)

# calculate coefficients
def a_n(n):
    return (2/l)*np.trapz(y*np.sin(n*np.pi*x/l), x)

# calculate u
def u(x, t, n_terms):
    u_xt = np.zeros_like(x)
    for n in range(0, n_terms+1):
        u_xt += a_n(n)*np.sin(n*np.pi*x/l)*np.cos(n*np.pi*c*t/l)
    return u_xt


n_values=[1,5,50,100]
for n_terms in n_values:
    plt.figure(figsize=(10,6))
    for t_i in np.linspace(0,2,5):
         plt.plot(x, u(x, t_i, n_terms), label=f't={t_i}')
    plt.title(f'Solution u(x,t) with n={n_terms}')
    plt.xlabel('x')
    plt.ylabel('u(x,t)')
    plt.legend()
    plt.grid(True)
    plt.show()
```
