## Coeficientes variables en la ecuacion hiperbolica.
Empalme con la clase anterior.

$$a(x,t) u_{tt} + 2 b(x,t) u_{tx} + c(x,t) u_{xx} + d(x,t) u_t + e(x,t) u_x + f(x,t) u = g(x,t)  \quad , \quad x \in \mathbb{R} \quad , \quad t>0. \tag{4a}$$



Lo que vamos a hacer es cambiar de coordenadas. Llevar a la forma **canonica**.

La nueva ecuacion queda

$$b^*(\xi, \eta) u_{\xi \eta} + d^*(\xi, \eta) u_\xi + e^*(\xi, \eta) u_\eta + f^*(\xi, \eta) u = g^*(\xi, \eta) .$$
La idea es pues eliminlar los terminos con coefficientes $a(\xi, \eta)$ y $c(\xi, \eta)$

La transformacion
$$\xi = \xi(x,t) \quad , \quad \eta = \eta(x, t) \tag{5} $$

Para este cambi de coordenadas sea regular (se pueda hacer sin infinitos)
el jacobiano debe ser distinto de cero.

$$ J = \begin{vmatrix} \xi_t & \xi_x \\ \eta_t & \eta_x \end{vmatrix} \ne 0  .$$


Comencemos la busqueda de todos los coeficientes estrella, al insertar (5) en
 .


usando la regla de la cadena para derivadas parciales
$$u_t = u_\xi \xi_t + u_\eta \eta_t $$
$$u_x = u_\xi \xi_x + u_\eta \eta_x \tag{6} $$

Ahora aplicando la regla de la cadena de nuevo.

\begin{eqnarray}
u_{tx} &=& (u_t)_x  = (u_\xi \xi_t + u_\eta \eta_t)_x \\
&=& (u_\xi \xi_t )_x + (u_\eta \eta_t)_x \\
&=& (u_\xi)_x \xi_t + u_\xi \xi_{tx} + (u_\eta)_x \eta_t + u_\eta \eta_{tx}
\end{eqnarray}
Recordemos ecuacion (6)

$$u_{tx} = (u_{\xi \xi} \xi_x + u_{\eta \xi}\eta_x) \xi_t + u_{\xi} \xi_{tx} + (u_{\eta \xi} \xi_x + u_{\eta \eta} \eta_x) \eta_t + u_{\eta} \eta_{tx} .$$


Escribo la ecuacion anterior con otro orden

$$u_{tx} = u_{\xi \xi} \xi_t \xi_x + u_{\xi \eta} (\xi_t \eta_x + \xi_x \eta_t) + u_{\eta \eta} \eta_t \eta_x + u_\xi \xi_{tx} + u_\eta \eta_{tx} .$$

Lo que sigue es $u_{tt}$ y $u_{xx}$, podemos copiar/pegar/editar, combiando primero $x$ por $t$ y despues $t$ por $x$.

$$u_{tt} = u_{\xi \xi} \xi_t^2 + 2 u_{\xi \eta} \xi_t \eta_x  + u_{\eta \eta} \eta_t \eta_t + u_\xi \xi_{tt} + u_\eta \eta_{tt} .$$


$$u_{xx} = u_{\xi \xi} \xi_x \xi_x + 2 u_{\xi \eta} \xi_x \eta_x  + u_{\eta \eta} \eta_x \eta_x + u_\xi \xi_{xx} + u_\eta \eta_{xx} .$$

Insertamos todo esto en  (4a) que es
$$a(x,t) u_{tt} + 2 b(x,t) u_{tx} + c(x,t) _{xx} + d(x,t) u_t + e(x,t) u_x + f(x,t) u = g(x,t)  \quad , \quad x \in \mathbb{R} \quad , \quad t>0. \tag{4a}$$


y encontramos


$$ a^* u_{\xi \xi} + 2 b^* u_{\xi \eta}  + c^* u_{\eta \eta} + d^* u_{\xi}  + e^* u_\eta  + f^* u = g^*. \tag{7}$$
donde





\begin{eqnarray}
a^* &=& a \xi_t^2 + 2 b \xi_t \xi_x + c \xi_x^2 \\
b^* &=& a \xi_t \eta_t + b ( \xi_t \eta_x + \xi_x \eta_t) + c \xi_x \eta_x \\
c^* &=& a \eta_t^2 + 2 b \eta_t \eta_x + c \eta_x^2 \\
d^* &=& a \xi_{tt} + b  \xi_{tx} + c \xi_{xx} +  d \xi_t + e \xi_x \\
e^* &=& a \eta_{tt} + b \eta_{tx} + c \eta_{xx} + d \eta_t + e \eta_x \\
f^* &=& f \\
g^* &=& g \tag{8}
\end{eqnarray}

Calculemos el nuevo determinante, en terminos de los coeficientes
$b^2, ac, a^2, c^2, ab, bc$

\begin{eqnarray}
D^* &=& (b^*)^2 - (a^*)(c^*) \\
&=& b^2[ (\xi_t \eta_x + \xi_x \eta_t)^2 - 4 \xi_t \xi_x \eta_t \eta_x] - ac[ \xi_t^2 \xi_x^2 + \eta_t^2 \xi_x^2 - 2 \xi_t \eta_t \xi_x \eta_x] \\
&& a^2 [\xi_t^2 \eta_t^2 - \xi_t^2 \eta_t^2] + c^2[ \xi_x^2 \eta_x^2 - \xi_x^2 \eta_x^2] + ab [ 2 \xi_t \eta_t ( \xi_t \eta_x + \xi_x \eta_t) - 2 \xi_t^2 \eta_t \eta_x -2 \eta_t^2 \xi_t \xi_x] + bc[ 2 \xi_x \eta_x( \xi_t \eta_x + \xi_x \eta_t) - 2 \eta_x^2 \xi_t \xi_x - 2 \xi_x^2 \eta_t \eta_x]
\\ &=&
b^2 ( (\xi_t \eta_x - \xi_x \eta_t)^2 - ac( \xi_t \eta_x - \xi_x \eta_t)^2 + a^2 0 + c^2 0 + ab 0  + bc 0  \\
&=& (b^2 - ac) J^2
\end{eqnarray}

La conclusion es que los distriminantes preservan el signo ante el cambio de trabnsformadas. Es decir son **invariantes bajo el cambio de coordenadas**.

La idea detras delos cambios de variables es hacer 0 los coeficientes de $a^* = c^*=0$, ecribamos las ecuaciones $a^*$ como

$$a \zeta_t^2 + 2 b \zeta_t \zeta_x + c \zeta_x^2 = 0 .$$
Donde $\zeta$ puede ser $\xi, o \quad \eta$.
Dividmos esta ecuacion por $\zeta_x^2$ y obtenemos

$$a \left ( \frac{\zeta_t}{\zeta_x}  \right )^2 + 2 b \left ( \frac{\zeta_t}{\zeta_x}    \right ) + c = 0   \tag{1} .$$

Recuerde que la idea de las variables $\zeta$ es que sean caracteristicas, es decir que la solucion $u$ sea constante para estas variables.
Si $\zeta(x,t) = \text{constante}$, por la definicion de la **diferencial**


$$ d \zeta = \zeta_t dt + \zeta_x dx = 0 .$$
De aca que

$$ \frac{\zeta_t}{\zeta_x} = -\frac{dx }{dt} .$$
Reemplazamo esto en (1)


$$ a \left ( \frac{dx}{dt} \right )^2 - 2 b \left ( \frac{dx}{dt}  \right ) + c = 0 .$$


Usando la formula cuadratica

$$\frac{dx}{dt} = \lambda_{1,2} = \frac{-b \pm \sqrt{b^2 - ac}}{a} .$$

Las soluciones son las caracteristicas que deben ser integradas.

## Ecuacion (de onda 1D) con fuente (no homogenea)
La forma de la ecuacion de onda con fuente es


$$ Lu = a(x,t) u_{tt} + 2 b(x,t) u_{xt} + c(x,t)u_{xx} + d(x,t) u_t + e(x,t) u_x + f(x,t) u = g(x,t).$$
donde ahora $g(x,t)$ es la fuente
La ecuacion $Lu=0$ es homogenea y si $u_1, u_2$ son soluciones entonces
$k_1 u_1 + k_2 u_2$ tambien lo es, por que $L$ es lineal.





Se sospecha que la fuente se integra dos veces, por que si
en el caso mas simple


$$ \frac{\partial^2  u}{\partial t \partial x} = g(x,t) .$$

Consideremos el siguiente IVP


\begin{eqnarray}
\begin{cases}
u_{tt} - c^2 u_{xx} = F(x,t) \quad , -\infty < x < \infty \quad , t>0 \tag{10} \\
u(x,0) = \varphi(x)  \quad , \quad  \left . \frac{\partial u(x,t)}{\partial t} \right . |_{t=0} = \psi(x)
\end{cases}
\end{eqnarray}


Para poder integrar este sistema necesitamos usar el teorema de Green.
Sean $P =P(x,t)$, $Q=Q(x,t)$ con dervadas parciales continuas en una camino suave orientado contrario a las manecillas del relog.

El teorema de Green dice

$$\iint \limits_{\Omega} [ Q_x(x,t) - P_t(x,t)  dx dt = \oint_{\Gamma} P(x,t) + Q(x,t) dt .$$

Integramos la ecuacion de onda

$$-\iint \limits_{\triangle} F(x,t) dx dt  = \iint \limits_{\triangle} (c^2 u_{xx} - u_{tt}) dx dt $$



donde $Q=c^2 u_x$, $P=u_t$
$$-\iint \limits_{\triangle} F(x,t) dx dt = \oint_{\Gamma} u_t dx + c^2 u_x dt .= \int_B + \int_R + \int_L u_t dx + c^2 u_x dt$$

Comencemos con la base. En la base $dt=0$

$$\int_B u_t dx + c^2 u_x dt = \int_{x_0 - c t_0}^{x_0+ct_0} u_t dx = \int_{x_0 - ct_0}^{x_0 + ct_0} \psi(x) dx. \tag{11} $$
A la derecha en $R$ tenemos que la caracteristica es $x_0+ct_0 = x_0$, entonces $dx_0=-c dt_0$. Hacemos esta sustitucion


$$\int_R u_t dx + c^2 u_x dt = -c \int_R u_t dt + u_x dx = -c \int du = -c[u(x_0,t_0) - u(x_0+ c t_0, 0)] = -c [ u(x_0, t_0) - \varphi(x_0 + c t_0). \tag{12} $$

Lo mismo analisis para el lado izquierdo
pero ahora $x-ct = x_0 - c t_0$ y $dx = cdt

$$\int_L u_t dx + c^2 u_x dt = c \int_R u_t dt + u_x dx = c \int du = -c[u(x_0,t_0) - u(x_0+ c t_0, 0)] = c [ \varphi(x_0 - c t_0) - u(x_0, t_0). \tag{13}$$

Sumando (11), (12), (13)


$$-\iint \limits_{\triangle} F(x,t) dx dt = \int_{x_0- c t_0)}^{x_0 + c t_0} \psi(s) ds + c[ \varphi(x_0 - c t_0) + \varphi(x_0 + c t_0) - 2 u(x_0, t_0) .$$

Despejando $u(x_0, t_0)$o

$$u(x_0, t_0) = \frac{\varphi(x_0 + c t_0) + \varphi(x_0 - c t_0)}{2} + \frac{1}{2c} \int_{x_0 - c t_0}^{x_0 + c t_0} \varphi(x) dx  + \frac{1}{2c} \iint \limits_{\triangle} F(x,t) dx dt .$$

Para un punto arbitrario $(x,y)$

$$u(x, t) = \frac{\varphi(x + c t) + \varphi(x - c t)}{2} + \frac{1}{2c} \int_{x - c t}^{x + c t} \varphi(x) dx  + \frac{1}{2c} \iint \limits_{\triangle} F(x,t) dx dt .$$
