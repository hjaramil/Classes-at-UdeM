La definicion general de una PDE de segundo orden

$$ F(x_1, x_2, \cdots, x_n, u, u_{x_1}, u_{x_2}, \cdots u_{x_{n}}, u_{x_{11}}, u_{x_{12}}, \cdots u_{x_{nn}})=0.$$

Definamos los terminos aca

* $u=u(x_1, x_2, \cdots, x_n)$ es una funcion real en las variables $x_1, x_2, \cdots, x_n$. U es un **campo escalar**.
*
$$ u_{x_i} = \frac{\partial u}{\partial x_i}, x_i \in \mathbb{R} \quad , i=1,2, \cdots, n $$

* Es necesario que por lo menos un $u_{x_i}$ este explicitamente para que sea por lo menos de primero orden en $F$ y si no no es una ecuacion diferencial. Es algebraica.
* Es necesario que exista en $F$ por lo menos un $u_{x_{ij}}$ donde

$$ u_{x_{ij}} = \frac{\partial^2 u}{\partial x_i \partial x_j} .$$
para que sea de segundo orden.

Nota: lo de **lineal** esta en las potencias (deben ser 1)lo de orden esta en las maximas derivadas, sin 2 , 3, 4.  En este curso hacemos maximo hasta 2.



## Motivacion. Derivacion de la ecuacion de onda.

A lo largo del horizontal hay equilibrio (no hay movimiento) entonces

$$T(x) \cos \alpha = T(x+ \Delta x) \cos \beta = T  \tag{1} .$$

A lo largo de la vertical usando la **segunda ley de Newton**.

$$-T(x) \sin \alpha + T(x + \Delta x) \sin \beta = m\frac{\partial^2 u}{\partial t^2}  \tag{2} .$$

Llamemos $\rho$ a la **densidad lineal**, de modo que masa es
$m= \rho \Delta s$.

$$ \Delta s = \Delta x^2 + \Delta u^2 .$$



De la ecuacion (1) despejando $T$


$$ T(x) = \frac{T}{\cos \alpha} \quad \text{y} \quad T(x + \Delta x) = \frac{T}{\cos \beta}  \tag{3} .$$

Escribamos la ecuacion (2) usando la (3)

$$-T \tan \alpha + T \tan \beta = \rho \Delta s \frac{\partial^2 u}{\partial t^2}  \tag{4}.$$

Ahora

$$\tan \alpha = \frac{\partial u}{\partial x}(x,t) \quad , \quad \tan \beta = \frac{\partial u}{\partial x} (x+ \Delta x,t)  .$$
Dividamos por $\Delta x$ ambos terminos


Escribimos la ecuacion (4) en estos terminos y la dividimos por $T \Delta x$


$$ \frac{\frac{\partial u}{\partial x} (x+ \Delta x)- \frac{\partial u}{\partial x}(x,t)}{\Delta x} = \frac{\rho \Delta s}{T \Delta x}  \frac{\partial^2 u}{\partial^2 t}. \tag{4b}$$

Que pasa con $\Delta s$

Recuerden de calculos


$$\sqrt{1 + \epsilon^2} \approx 1 + \frac12 \epsilon^2 \tag{5} .$$
En este caso vamos a ver lo siguiente


$$ \Delta s = \sqrt{ \Delta x^2 + \Delta u^2} $$
$$ \frac{\Delta s}{\Delta x} = \sqrt{ \frac{\Delta x^2}{\Delta x^2} + \frac{\Delta u^2}{\Delta x^2}} = \sqrt{1  +\frac{\Delta u^2}{\Delta x^2}} $$
Tomando el limite cuando $\Delta x \to 0$,


$$ \frac{ds}{dx} = \sqrt{1 + \left ( \frac{\partial u}{\partial x} \right )^2}. \tag{6}$$

Tomemos el limite en la Ecuacion (4b) cuando $\Delta x \to 0$ y obtenemos



$$ \lim_{\Delta x\to 0} \frac{\frac{\partial u}{\partial x} (x+ \Delta x,t)- \frac{\partial u}{\partial x}(x,t)}{\Delta x} = \lim_{\Delta x \to }  \frac{\rho \Delta s}{T \Delta x}  \frac{\partial^2 u}{\partial^2 t}$$

$$\frac{\partial^2 u}{ \partial x^2} =  \frac{\rho}{T} \sqrt{1 + \left ( \frac{\partial u}{\partial x} \right)^2 } \frac{\partial^2 u}{\partial^2 t}. $$

Hacemos la siguiente linealizacion $\partial u/\partial x \approx 0$ y

$$\frac{\partial^2 u}{ \partial x^2} =  \frac{\rho}{T}  \frac{\partial^2 u}{\partial^2 t}. $$

Esta es le ecuacion de onda en una cuerda.

Existe una Ecuacion llamada de Laplace


$$\nabla^2 u(x,y) = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0 .$$

Verficamos usando `SymPy` que $u(x,y) = \log(x^2 + y^2)$ satisface esta ecuacion.


```python
import sympy as sp
from sympy import pprint, init_printing

# Enable pretty printing
init_printing()

# definamos los simbolos
x,y = sp.symbols('x y')

# definamos la funcion u(x,y)
u = sp.log(x**2 + y**2)

# vector de segundas derivadas parciales
uu = [sp.diff(u, var, 2) for var in [x,y]]
display(uu)
```

```python
# calculamos el laplaciano
laplacian_u = sum( uu[i] for i in range(2))

# display a la funcion y su laplaciano
display("function u", u)
display("Laplacian ", laplacian_u)
```

```python
# simplifiquemos
display("Laplacian simplified", sp.simplify(laplacian_u))
```

En general una PDE de primer orden se define


$$F(x_1, x_2, \cdots, x_n, u, u_{x_1}, u_{x_2}, \cdots, u_{x_n}) = 0 .$$

Inicialmente, para introducir el **metodo de las caracteristicas** estudiamos la siguiente ecuacion


$$ \frac{\partial u(t,x)}{\partial t} + \frac{u(t,x)}{\partial x} = 0  \tag{2.0.2}.$$

En este ejemplo $x_1=t$, $x_2=x$ y $F = u_t + u_x$.

## Econtrando las caracteristicas.

Definamos el vector


$$\bma = \binom{1}{1} .$$


Nota todo debe ser $u(t,x)$ no $u(x,t)$ cambiar global.
Ahora

$$ \bma \cdot \nabla u = 0.$$
El gradiente $\nabla u$ es perpendicular a las curvas de contorno.  (calculo en varias variables)

Ademas, tomando la derivada total con respecto a la longitud de arco $s$,
usando la **regla de la cadena**

Seleccionamos un contorno arbitrario y tomamos la derivada a lo largo de
este contorno con respecto a la longitud de archo $s$. Esta deriva es 0.
$$\frac{du}{ds} = \frac{\partial u}{\partial x} \frac{dx}{ds} + \frac{\partial u}{\partial t} \frac{dt}{ds}  = 0 .$$

Si llamamos $\bmb$


$$ \bmb = \begin{pmatrix} {\frac{dx}{ds}} \\{\frac{dt}{ds}} \end{pmatrix} .$$

Esto me dice que el vector $\bmb$ tambien es perpendicular al $\nabla u$
El gradiente $\nabla u$  es perpendicular al contorno, y $b$ perpendicular al gradiente. O sea que $b$ esta en la direccion del vector tangente $\bmt$

Esto quiere decir que si

$$\bmt = \begin{pmatrix}  \frac{dx}{ds}, \frac{dt}{ds} \end{pmatrix} .$$
y es paralelo $b$

Como tanto $\bma$ como $\bmb$ son perpendiculares al gradiente, entonces son paralelos

$$ \bmb = k \bma .$$
donde $k \in \mathbb{R}$.

Es decir

\begin{eqnarray*}
 \frac{dx}{ds} = k(1)  \\
 \frac{dt}{ds}  = k(1) \\
\end{eqnarray*}

Ahora podemos dividir estas dos ecuaciones


$$\frac{\frac{dx}{ds}}{\frac{dt}{ds}} = \frac{dx}{ds} \frac{ds}{dt} = \frac{dx}{dt} = 1 .$$

Redujimos el problema de una ecuacion en derivadas parciales a una ecuacion ordinaria. En esto consiste el metodo de las caracteristicas.

Esta ecuacion $dx/dt=1$ es facil de integrar

$$ x = t + c$$

A estas graficas con $c$ cambiando se les llama **caracteristicas base**.

El significado de $dx/dt$ es velocidad (cuando $t$ es tiempo y $x$ es espacio).
En general la ecuacion de transporte se escribe como

$$ \frac{\partial u}{\partial t} + v \frac{\partial u}{\partial x} = 0 .$$

Vemos como el coeficiente $v$ representa velocidad cuando resuelva este problema.
