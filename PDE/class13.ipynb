## Funciones de Green.
### Motivacion
Recuerden del algebra lineal el problema

$$ Ax = b$$

En ODEs y PDEs existe un problema similar

$$ Lu = f$$
La ventaja de tener $A^{-1}$ es que yo puedo resolver no solo $Ax=b$, sino $Ax=y$ o cualquier "fuente" que usted se le antoje.

Por ejemplo suma $e_1, e_2, e_3, \cdots$ los vectores canonicos con $1$ en la posicion $i$ y 0 en las demas (para $e_i$.
y un conjunto de vectores $x_1, x_2, \cdots, x_n$ tales que

$$ A x_1 = e_1 $$
$$ A x_2 = e_2 $$
$$ A x_3 = e_3 $$
$$ \vdots $$
$$ A x_n = e_n $$


Esto sistema se puede escribir en forma matricial como

$$A[x_1 | x_2 | \cdots | x_n ] = [ e_1 | e_2 |   \cdots | e_n ] $$
En forma compacta esto es


$$AB = I  $$
Esto es
$$(A x_i)_j = \delta_{ij}

Cuando estiemos Green functions vamos a ver
$$ L (G( x, \xi) = \delta(x - \xi) $$
Esto es similar a $Lu=f$ y
la comparacion es la siguiente

$$ A \leftrightarrow L$$
$$ G \leftrightarrow A^{-1}$$
$$ i \leftrightarrow x$$
$$ j \leftrightarrow \xi $$

Como usamos la funcion de Green para resolver $Lu=f$?

Recordemos el sistema finito $Ax=b$. Como encontramos $x$?

$$ A^{-1}Ax = A^{-1} b $$
$$ x = A^{-1} b $$

De igual forma pasa con funciones de Green en espacios ininitodimensionales.

$Lu=f$,
$G=L^{-1}$

$$u = L^{-1} f = \langle G(x,\xi), f(\xi)   \rangle = \int_{-\infty}^\infty G(x, \xi) f(\xi) d \xi \tag{1} $$

Otra forma de ver el resultado (1) es la siguiente. **Superposicion**

Recuerden  (sifting property of the Dirac Delta)

$$f(x) = \int_{-\infty}^\infty f(\xi) \delta(x - \xi) d \xi \tag{2} $$
Ahora bien
$$ L G(x, \xi) = \delta(x - \xi) \tag{3} $$
Reemplazamos (3) en (2)

$$f(x) = \int_{-\infty}^\infty f(\xi) L[ G(x, \xi)] d\xi $$
como $Lu = f$ escribimos

$$Lu(x) = \int_{-\infty}^\infty f(\xi) L[ G(x, \xi)] d\xi $$
usamos linealidad.


$$Lu(x) = \int_{-\infty}^\infty F[ f(\xi) G(x, \xi)] d \xi  = L \left [ \int_{-\infty}^\infty f(\xi) G(x, \xi) d \xi   \right ] $$

Si $L^{-1}$ existe, podemos aplicarlo a ambos lados y obtenemos

$$u(x) =  \left [ \int_{-\infty}^\infty f(\xi) G(x, \xi) d \xi   \right ] $$


Este ejemplo ocurre con ecuaciones lineales ordinarias (ODE), pues hay una sola variable. Que pasa si en vez son parciales. Hay por lo menos dos variables.
Asumamos las variables $(x,t)$

La anologia es la misma. Al lado izquierdo en vez de $\delta(x-\xi)$ tendriamos

$$\delta(x-\xi) \delta(t - \tau) $$

La funcion de Green tendria cuatro arumentos

$$G(x, t; \xi , \tau) $$
En teneral si hay $n$ variables, debe haber $n$ deltas y $2n$ argumentos en la funcion de Green

La idea es igual a la anterior y el sistema que formulamos es

$$LG(x,t; \xi, \tau) = \delta(x-\xi) \delta(t - \tau) $$

De la misma forma si $Lu = f$ ,  $u = G^{-1} f$ esto es un producto interno


$$u(x,t) = \langle G(x,t,. ,. ), f \rangle = \int_{-\infty}^\infty \int_{-\infty}^t G(x, t, \xi, \tau) f(\xi, \tau) d \xi d \tau. $$

En este punto movtivamos las funciones de Green, mostramos como se usan para hallar soluciones al problema $Lu=f$ (ecuacion diferencial ordinaria o partcial)
pero no hemos describto **como hallarla** (algoritmos para encontrarla).

Existen los siguientes algoritmos:

* **Para dominios infinitos** : Se usa la **transformada de Fourier**.
* ** **Para dominios finitos** o **semi-infinitos** se usa la **Transformada de Laplace**.
* El **metodo de imagenes** (se usa mucho  en fisica). Tambien se usan ecuaciones integrales.

En esta clase solo discutimos los metodos de Fourier.

## Transformada de Fourier.

### Teoria de transformadas.
Asuma un operador $L$ en un espacio $S$ (funciones Lebesgue integrales $S=L^2[a,b]$.  Asumamos que el operador tiene un **espectro** discreto

$$L \varphi_n = \lambda_n \varphi_n  $$
Asumamos que las autofunciones $\varphi_n$ estan normalizadas y son ortogonales. Por ejemplo que el operador $L$ es SLP (autoadjunto) y que las autofunciones forman un conjunto **completo** en $S$. Entonces tenemos el siguiente para de ecuaciones


$$ \begin{cases} f(x) = \sum_{i=1}^\infty c_i \varphi_i(x)   \\ c_i = \langle f, \varphi_i \rangle  = \int_D f(x) \varphi(x) dx  \end{cases}$$

A esto se le llaman un "transform pair".

Tenemos lo siguiente

1. La primera ecuacion reprsenta la suma **sintesis** de todas las autofunciones $\varphi_i$ que constituyen el **espectro**.
2. La segunda extrae las componentes individuales del espectro **colores**. A este proceso se le llama **analisis**.

Vemos ejemplos

**Ejemplo 1**: Cuando resolvemos el sistema

\begin{eqnarray}
\left \{
    \begin{array}{cc}
    F'' + \lambda^2 F = 0 \\
    F(0) = F(\ell) = 0.
    \end{array}
    \right .
\end{eqnarray}
con $x \in [0 , \ell]$


Encontramos

$$\lambda_n = \frac{n \pi}{\ell} $$
$$ F_n = \frac{2}{\ell } \sin \frac{n \pi x}{\ell}$$

El conjunto de funcion $\{ F_n \}$ es ortonormal y si la condicion inicial es $\varphi(x)$ tenemos

$$ \varphi(x) = \sum_{i=1}^{\infty} c_i F_i $$
$$c_i = \langle \varphi(x), F_i \rangle = \int_0^{\ell} F_i(x) dx $$

mas especificamente

\begin{eqnarray}
\varphi(x) = \sum_{i=1}^{\infty} c_i \sqrt{2}{\ell} \sin \frac{n \pi x}{\ell} \\
c_i = \frac{2}{\ell} \int_0^{\ell} \sin \frac{n \pi}{\ell} \varphi(x)  dx
\end{eqnarray}

* **Ejemplo 2**:
*\begin{eqnarray}
\left \{
    \begin{array}{cc}
    F'' + \lambda^2 F = 0 \\
    F(-L) = F(L) \quad , F'(-L)=F(L)
    \end{array}
    \right .
\end{eqnarray}
con $x \in [-L , L]$.

Ya sabemos que las soluciones de la ecuacion diferencial (ignorar los bordes) son
$$\mathrm{e}^{\mathrm{i} \lambda x}  \quad , \quad \mathrm{e}^{-\mathrm{i} \lambda x } $$

Cualquier combinacion lineal de estas tambien es solucion

$$F(x) =C_1 \mathrm{e}^{\mathrm{i} \lambda x}  + C_2 \mathrm{e}^{-\mathrm{i} \lambda x} $$

Apliquemos la condicion de fronteraa $F(-L) = F(L)$

$$C_1 \mathrm{e}^{- \mathrm{i} \lambda L} + C_2 \mathrm{e}^{\mathrm{i} \lambda L} =  C_1 \mathrm{e}^{\mathrm{i} \lambda L} + C_2 \mathrm{e}^{-\mathrm{i} \lambda L}$$
Simplifiquemos

$$-2 \mathrm{i} C_1 \sin \lambda L = -2 \mathrm{i} C_2 \sin \lambda L $$
$$2 \mathrm{i}(C_2 - C_1) \sin \lambda L = 0 $$

Ahora para $F'(-L) = F'(L)$
$$\mathrm{i} \lambda C_1 \mathrm{e}^{- \mathrm{i} \lambda L} - \mathrm{i} \lambda C_2 \mathrm{e}^{\mathrm{i} \lambda L} =  C_1  \mathrm{i} \lambda \mathrm{e}^{\mathrm{i} \lambda L} - \mathrm{i} \lambda  C_2 \mathrm{e}^{-\mathrm{i} \lambda L}$$


De forma que
$$-2 \mathrm{i} C_1 \sin \lambda L = 2 \mathrm{i} C_2 \sin \lambda L $$

$$\lambda(C_1 + C_2) \sin \lambda L = 0 $$
Tiene que pasar que $\sin \lambda L=0$ o sea que $\lambda L = n \pi$,
y
$$\lambda_n = \frac{n \pi}{L} $$
son los autovalores del operador (con condiciones de frontera).

Themos entonces que


$$F(x) = C_1  \mathrm{e}^{\mathrm{i} \frac{n \pi}{L} x} + C_2 \mathrm{e}^{-\mathrm{i} \frac{n \pi}{L} x}. $$


Esta suma se puede escribir como

$$F(x) = \sum_{n=-\infty}^\infty a_n \mathrm{e}^{-\mathrm{i} \frac{n \pi}{L} x} i \tag{5} $$

Veamos que las autofunciones $\{ \mathrm{e}^{\mathrm{i} \frac{n \pi}{L} x} \}$
son ortogonales.
Veamos, asuma $n \ne m$

\begin{eqnarray}
\langle \mathrm{e}^{\mathrm{i} \frac{n \pi}{L} x }, \mathrm{e}^{\mathrm{i} \frac{m \pi}{L} x } \rangle  &=& \int_{-L}^L \mathrm{e}^{\mathrm{i}(n -m) \frac{\pi}{L} x} dx \\
&=& \left . \frac{L}{\mathrm{i}(n -m) \pi} \mathrm{e}^{\mathrm{i}(n - m) \frac{\pi}{L} x} \right |_{-L}^L \\
&=& \frac{L}{\mathrm{i}(n-m) \pi} \left [  \mathrm{e}^{\mathrm{i}(n-m)\pi }- \mathrm{e}^{-\mathrm{i}(n-m) \pi}  \right ]  \\
&=& \frac{2 \mathrm{i}}{\mathrm{i}(n-m) \pi} \sin( (n-m) \pi) \
&=& 0
\end{eqnarray}
De forma que las autofunciones exponenciales son ortogonales.

Hallemos la norma


\begin{eqnarray}
\langle \mathrm{e}^{\mathrm{i} \frac{n \pi}{L} x }, \mathrm{e}^{\mathrm{i} \frac{n \pi}{L} x } \rangle  &=& \int_{-L}^L \mathrm{e}^{\mathrm{i}(n -n) \frac{\pi}{L} x} dx \\
&=& \int_{-L}^L dx \\
&=& 2L
\end{eqnarray}

\begin{eqnarray}
\langle \mathrm{e}^{\mathrm{i} \frac{n \pi}{L} x }, \mathrm{e}^{\mathrm{i} \frac{m \pi}{L} x }  \rangle  =  \delta_{nm} 2 L \
\end{eqnarray}
Hagamos la projeccion de $F(x)$ sobre las autofunciones

$$ \langle F(x) , \mathrm{e}^{\mathrm{i} \frac{k \pi}{L} x } \rangle = \sum_{n=-\infty}^{\infty} a_n \langle \mathrm{e}^{\mathrm{i} \frac{n \pi x}{L}} ,
\mathrm{e}^{\mathrm{i} \frac{k \pi x}{L}}\rangle = \sum_{n=-\infty}^{\infty} a_n 2 \delta_{nk} L = 2 a_k L     $$
de forma que

$$ a_n = \frac{1}{2L} \int_{-L}^L F(x) \mathrm{e}^{- \mathrm{i} \frac{n \pi x}{L}} dx \tag{6} $$

## La transformada de **Fourier**
De las ecuaciones (5) y (6) obtenemos el par (transformada)


$$F(x) = \sum_{n=-\infty}^\infty a_n \mathrm{e}^{-\mathrm{i} \frac{n \pi}{L} x}  $$

$$ a_n = \frac{1}{2L} \int_{-L}^L F(x) \mathrm{e}^{- \mathrm{i} \frac{n \pi x}{L}} dx. $$

Para hallar la transforma de Fourier vamos a sumir $L \to \infty$.
Hagamos un cambio  de variable

$$k(n , L) = \frac{n \pi}{L} $$
Cuando $L \to \infty$, $dk=k(n+1)-k(n-1)=\pi/L$.

Ahora

$$dn \to \frac{L dk}{\pi} $$
Queda lo siguiente

$$\sum_{-\infty}^\infty \to \int_{-\infty}^\infty \frac{L}{\pi} dk $$

$F(x)$ se convierte en

$$F(x) = \frac{L}{\pi} \int_{-\infty}^{\infty} a_k \mathrm{e}^{\mathrm{i}k x} dk$$

Renombremos el coeficiente

$$\frac{a_k L}{\pi} = \frac{1 L}{2 L \pi} \int_{-\infty}^{\infty} F(x) \mathrm{e}^{-\mathrm{i} k x} dx  $$

Queda el siguiente par
$$\hat{F}(k)= \frac{1}{2 \pi} \int_{-\infty}^\infty F(x) \mathrm{e}^{-\mathrm{i} k x } dx $$
$$F(x) = \int_{-\infty}^\infty \hat{F}(k) \mathrm{e}^{\mathrm{i} k x} dk $$

A este par se le llama el **Fourier Transform Pair**

Algunas notas:

* **Espacio**: Los dominios son $F(x)$ y $\hat{F}(k) $
* **Tiempo**: Los dominios son $F(t)$ y $\hat{F}(\omega) $
* $k$ frecuencia espacial
$\omega$ frecuencia temporal
* Si estuvieramos en la variable tiempo hubieramos obtenido el par de transormadas.
$$\hat{F}(\omega)= \frac{1}{2 \pi} \int_{-\infty}^\infty F(t) \mathrm{e}^{\mathrm{i} k t } dt $$
$$F(t) = \int_{-\infty}^\infty \hat{F}(\omega) \mathrm{e}^{-\mathrm{i} \omega t} d \omega $$


* No hay una **unica convencion**. A veces el $1/2 \pi$ esta a un lado (tiempo), o a veces al otro (frecuencia). A veces el signo del tiempo es + y el de frecuencia "-" a veces al contrario. Lo importante es que si se usa una convencion para convertir los datos de tiempo a frecuencia, se **respete** esta convencion cuando se tome la transformada inversa. Ver
[esta pagina](https://civil.poriyaan.in/topic/fourier-transform-pair-40331/#google_vignette) para las distintas convenciones.





Para interpretar los simbolos $k, \omega$. Pensemos en una solucion monocromatica.

$$u(x,t, k, \omega) = \mathrm{e}^{\mathrm{i}( k x \pm \omega t)} $$
Sabemos que esto se puede escribir como senos y cosenos, y como el coseno es un seno con un corrimiento de fase ($\cos(x + \pi/2)= - \sin x)$
Podemos escribir

$$u(x,t,k,\omega) = \cos( kx \pm \omega t) = \cos k \left ( x \pm \frac{\omega}{k} t  \right ) \tag{7} $$

Si se inserta esta ecuacion en la ecuacion de onda $u_{tt} - \frac{1}{c^2}u_{xx} =0$
obtiene

$$c = \frac{\omega}{k} $$
De acuerdo a esto escribrimos

$$u(x,t,k,c) = \cos k ( x \pm ct) $$

Podriamos haber factorizado $\omega$ de (7)

$$u(x,t,k, \omega_ = \cos (kx \pm \omega t) = \cos \omega \left ( t - \frac{k}{\omega} x \right ) = \cos \omega \left ( t - \frac{x}{c} \right )$$

* Mas interpretaciones:

$$\omega = 2 \pi f = \frac{2 \pi}{T} $$
$$ k = \frac{2 \pi}{\lambda} = \frac{2\pi}{c T} $$
donde
* $T$: **periodo**: Tiempo en que la onda recorre una distancia $\lambda$.
* $\lambda$: longitud de onda. Cada onda monocromatica tiene una distance donde se repite.
* $f$ frecuencia en Hertz. $1/T$
* $k$ numero de onda (frecuencia espacial).

* Notacion adicional.
$ \hat{F}(k)$ se llama la **transformada directa**

$F(x)$ se llama la **transformada inversa**

Ahora otra notacion

$$ \hat{F}(k) = \mathcal{F} (F (x)) $$
$$ F(x) = \mathcal{F}^{-1} (F (k)) $$
Claramente

$$ \mathcal{F}^{-1}(\mathcal{F}(F(x))) = F(x) $$
$$ \mathcal{F}(\mathcal{F}^{-1}(F(k))) = F(k) $$

* **Ejemplo 3**:
Asumimos una senhal con 4 frecuencias $(0, 5,10,20)$, tres fases $(0,\pi/4,\pi/2)$ y 4 amplitudes $(0, 1, 0.5, 0.2)$.



```python
import numpy as np
import matplotlib.pyplot as plt

# parameters for the signal
Fs=500 # sampling frequency
T = 1/Fs # periodo
L=1000 # longitud de la senahal
t = np.arange(0,L )*T # vector de tiempos


# definamos las componentes monocromaticas
frequencies = [5, 10, 20]
amplitudes = [1, 0.5, 0.2]
phases = [0, np.pi/4, np.pi/2]

# generate monochromatic wavelets
individual_waves = []
summed_signal = np.zeros(L)

for freq, amp, phase in zip(frequencies, amplitudes, phases):
    wave = amp * np.sin(2 * np.pi * freq * t + phase)
    individual_waves.append(wave)
    summed_signal += wave

# graficamos las ondiculas monocromaticas
plt.figure(figsize=(12, 12))
for i, wave in enumerate(individual_waves):
    plt.subplot(6, 1, i + 1)
    plt.plot(t, wave, label=f'Frequency: {frequencies[i]} Hz')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title(f'Monocromatica {i +1} with frequency {frequencies[i]} Hz')
    plt.grid(True)
    plt.legend()

# plot the synthesis of the signal by adding the wavelets
plt.subplot(6, 1, 4)
plt.plot(t, summed_signal, color='green' , label='Sum of Monochromatics')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Synthesis of the Signal')
plt.grid(True)
plt.legend()

# plot the spikes to indicate the frequency componets
plt.subplot(6,1,5)
plt.stem(frequencies, amplitudes, use_line_collection=True, basefmt=" ", linefmt="r-")
plt.title("Ampltiude spectrum of the signal")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Frequency Components')
plt.grid(True)

plt.subplot(6,1,6)
plt.stem(frequencies, phases, use_line_collection=True, basefmt=" ", linefmt="r-")
plt.title("phase spectrum of the signal")
plt.xlabel('Frequency (Hz)')
plt.ylabel('phase')
plt.title('Frequency Components')
plt.grid(True)


plt.tight_layout()
plt.show()
```

### Datos adicionales.
La transformada de Fourier produce numeros complejos. El el espectro de amplitude se calcula con la formula

$$| \hat{f}(k) | = \sqrt{ \text{Re}(\hat{F}(k))^2 + \text{Im}(\hat{F}(k))^2} $$

La phase es la tangente inversa de la parte imaginaria sobre la real

$$\text{fase}= \arctan \frac{\text{Im}(\hat{F}(k))}{\text{Re}(\hat{F}(k))} $$

## Propiedades de la transformada de Fourier
### Transformada de Fourier de un "shift" (corrimiento)
Asuma que $g(x) = F(x - s)$, donde $s \in \mathbb{R}$ es un corrimiento
entonces

$$\mathcal{F}[g(x)] = \frac{1}{2 \pi} \int_{-\infty}^{\infty} g(x) \mathrm{e}^{-\mathrm{i} k x} dx = \frac{1}{2 \pi} \int_{-\infty}^{\infty} f(x-s) \mathrm{e}^{-\mathrm{i} k x} dx $$
Hacemos cambio de variable $u=x-s$ entonces $du=dx$, $x=s+u$ o sea

$$\mathcal{F}[f(x-s)] = \mathrm{e}^{-\mathrm{i} s} \frac{1}{2 \pi} \int_{-\infty}^{\infty} f(u) \mathrm{e}^{-\mathrm{i} k u} = \mathrm{e}^{-\mathrm{i} s} \mathcal{F}[f(x)]$$

### Transformada de Fourier del Delta de Dirac
$$\mathcal{F}(\delta(x)) = \frac{1}{2 \pi} \int_{-\infty}^{\infty} \delta(x) \mathrm{e}^{-\mathrm{i} k x} dx = \frac{1}{2 \pi}$$
