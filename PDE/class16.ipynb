$$u_t - \alpha( u_{xx} + u_{yy}) = 0 \quad \text{PDE sin fuente} \tag{1} $$
$$u(0,y,t) = u(L_x, y, t)= u(x, 0,t) = u(x, L_y,t) = 0

$$u(x,y,0) =  \varphi(x,y) $$


Asumimos
$$u(x,y,t) = X(x) Y(y) T(t) $$
Si sustimos esto en la ecuacion (1)

$$X(x) Y(y) T'(t) = \alpha ( X''(x) Y(y) T(t) + X(x) Y''(y) T(t) $$
Dividimos todo por $XYT$

$$\frac{T'(t)}{\alpha T(t)} = \frac{X''(x)}{X(x)}  + \frac{Y''(y)}{Y(y)} $$

Como siempre esto es cierto solo si los lados izquierdo y derecho son constantes. El lado derecho tiene dos partes, estas son constantes.

Empecemos con el lado izquierdo


$$\frac{T'(t)}{\alpha T(t)} = - \lambda^2 $$
Por que negativo?

La solucion es exponencial (por eso negativa para que decaiga)

De igual forma

$$\frac{X''(x)}{X(x)}  = - \mu^2 \quad , \quad \frac{Y''(y)}{Y(y)}=-\gamma^2 $$
donde
$$\mu^2 + \gamma^2 = \lambda^2 $$

Podemos escribir el sistema como

$$\frac{X''(x)}{X(x)}  = - \mu^2 \quad , \quad \frac{Y''(y)}{Y(y)}=
\mu^2 -  \lambda^2 $$
para poner todo en terminos solo de $\mu, \lambda$.

Resolvemos la ecuacion en $X$

$$X''(x) + \mu^2 X(x) = 0  $$
Sabemos que las soluciones son senos y cosenos

$$X(x) = C_1 \cos \mu x + C_2 \sin \mu x $$

Por condiciones iniciales $u(0,y,t)=X(0)Y(y)T(t)=0$ como no queremos la
trivial
$$X(0) = C_1 = 0 $$
De forma

$$X(x) = C_2 \sin \mu x $$
La condicion en $L_x$ da

$$X(L_x) = C_2 \sin \mu L_x = 0 $$

como no queremos $C_2 = 0$,  quiere decir $\mu L_x = m \pi$ y

$$\mu = \mu_m =\frac{m \pi}{L_x} $$
Ahora la ecuacion en
$$Y''(y) + (\lambda^2 - \mu^2)  Y(y) = 0  $$
Sabemos que las soluciones son senos y cosenos
con $\gamma^2 = \lambda^2 - \mu^2$

$$Y(y) = C_3 \cos \gamma x + C_4 \sin \gamma x $$
Por condiciones iniciales $u(x,0,t)=X(x)Y(0)T(t)=0$ como no queremos la
trivial
$$Y(0) = C_3 = 0 $$
De forma

$$Y(y) = C_4 \sin \gamma x $$
La condicion en $L_y$ da

$$Y(L_y) = C_4 \sin \gamma L_y = 0 $$
como no queremos $C_4 = 0$,  quiere decir $\gamma L_y = n \pi$ y

$$\gamma = \gamma_n =\frac{n \pi}{L_y} $$


Con la relacion
$$\lambda_{mn}^2 = \left ( \frac{n \pi}{L_y}  \right )^2 + \left (\frac{m \pi}{L_x} \right )^2 \tag{2} $$


Regresemos a la ecuacion en tiempo

$$\frac{T(t)}{\alpha T(t)} = - \lambda^2 $$

Esto tiene soluciones exponenciales. La solucion es

$$T(t) = C_5 \mathrm{e}^{-\alpha \lambda^2 t} $$
Remplazamos (2) aca

$$T(t) = C_5 \mathrm{e}^{-\alpha \left [  \left ( \frac{n \pi}{L_y} \right )^2 + \left ( \frac{m \pi}{L_x} \right )^2 \right ] t} $$

## Separacion de variables para Laplace en 3 dimensiones espaciales

$$\nabla^2 u = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} + \frac{\partial^2 u}{\partial z^2} =  0 \tag{3} $$

No pongamos condiciones de frontera por el momento.

$$u(x,y,z) = X(x) Y(y) Z(z) $$.

Si insertamos esto en  (3) encontramos que


$$
\frac{X''(x)}{X(x)} + \frac{Y''(y)}{Y(y)} + \frac{Z''(z)}{Z(z)} = 0
$$

Vamos a las condiciones de frontera
Tenemos la ecuacio en una caja $[0, L_x] \times [0, L_y] \times [0, L_z]$
Las condiciones son

$$u(0, y, z) = u(L_x, y, z) = 0  $$
$$u(x, 0, z) = u(x, L_y, z) = 0  $$

La condicion en $z$ ( no puede ser cero pero la discutimos mas abajo)

Tenemos las ecuacion en $X,Y$

$$\frac{X''}{X(x)} = -\lambda^2 $$
$$\frac{Y''}{Y(y)} = -\mu^2 $$
y
$$\frac{Z''}{Z(z)} = \gamma^2 =\lambda^2 + \mu^2   $$

De la misma forma que hicimos en la ecuacion de onda

$$\lambda_m = \frac{m \pi}{L_x} \quad , X_m(x) = C_1 \sin \frac{m \pi x}{L_x} $$
$$\mu_n = \frac{n \pi}{L_y} \quad , Y_n(x) = C_2 \sin \frac{n \pi y}{L_y} $$

Ahora condiciones de frontera para $z$

$$u(x,y, 0 ) = 0 \quad , \quad u(x,y,L_z) = \varphi(x,y) $$

La ecuacion en $z$ es

$$Z''(z) - \lambda^2 Z(z) = 0 $$
La solucion de esto esta en terminos de exponenciales reales o senos y cosenos hiperbolicos. Es decir

$$Z(z) = C_3  \sinh \gamma z + C_4 \cosh \gamma z $$
Ahora como $Z(0)=0$ (asumiendo soluciones no triviales)
$Z(0)= C_4 = 0$
y
$$Z_{mn}(z) = C_3 \sinh \sqrt{ \left ( \frac{m \pi}{L_x}  \right )^2  + \left ( \frac{n \pi}{L_y} \right )^2} z $$

Hacemos superposicion

$$u(x,y,z) = \sum_{m=1}^{\infty} \sum_{n=1}^{\infty} a_{mn}
\sin \frac{m \pi x}{L_x} \sin \frac{n \pi y}{L_y}
\sinh \left ( \sqrt{ \left ( \frac{m \pi}{L_x}  \right )^2  + \left ( \frac{n \pi}{L_y} \right )^2} \right ) z$$

Projectamos la condicion inicial sobre las autofunciones

$$\varphi_{mn}(x,y) = \sin \frac{m \pi x}{L_x} \sin \frac{n \pi y}{L_y} $$
Son ortonales y ya sabemos que


$$\langle \varphi_{mn}, \varphi_{pq} \rangle = \delta_{mp} \delta_{nq} \frac{L_x L_y}{4} $$

La condicion de frontera en $z=L_z$ (top, arriba)

$$\varphi(x,y) = u(x,y,L_z) = \sum_{m=1}^{\infty} \sum_{n=1}^{\infty} a_{mn}
\sin \frac{m \pi x}{L_x} \sin \frac{n \pi y}{L_y}
\sinh \left ( \sqrt{ \left ( \frac{m \pi}{L_x}  \right )^2  + \left ( \frac{n \pi}{L_y} \right )^2} \right ) L_z$$


La proyeccion es:

\begin{eqnarray}
\langle \varphi(x,y), \varphi_{pq} \rangle &=&
\sum_{m=1}^{\infty} \sum_{n=1}^{\infty} a_{mn}
\left \langle \varphi_{mn} \sinh \sqrt{ \left ( \frac{m \pi}{L_x} \right )^2 + \left ( \frac{n \pi}{L_y} \right )^2}    L_z, \varphi_{pq} \right \rangle \\
&=&
\sum_{m=1}^{\infty} \sum_{n=1}^{\infty} a_{mn}
\sinh  \sqrt{ \left ( \frac{m \pi}{L_x} \right )^2 + \left ( \frac{n \pi}{L_y} \right )^2} L_2 \delta_{mp} \delta_{nq} \frac{L_x L_y}{4}  \\
&=& a_{pq} \frac{L_x L_y L_z}{4} \sinh  \sqrt{\left ( \frac{p \pi}{L_x} \right )^2 +
\left ( \frac{q \pi}{L_y}  \right)^2}
\end{eqnarray}

Se despeja $a_{pq}$ y llamamos $p \to m$, $q \to n$ y

$$a_{mn} = \frac{4}{ L_x L_y L_z \sinh \sqrt{ \left ( \frac{m \pi}{L_x} \right )^2 + \left ( \frac{n q}{L_y}\right )^2 }}
\int_0^{L_x} \int_0^{L_y} \varphi(x,y)
\sin \frac{m \pi x}{L_x} \sin \frac{ n \pi y}{L_y} dx dy. $$
