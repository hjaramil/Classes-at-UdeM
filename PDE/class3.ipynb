## Equaciones PDE de primer orden quasi-lineales.
La teoria de estas PDEs es practicamente la misma de las lineales. De hecho el libro de Bliestein no considera las lineales sino que se mete de una a estas. Yo considero las lineales por que gusta ir de abajo hacia arriba.

Vamos a repetir parte de lo que ya hicimos.

La ecuacion

$$a(x,y,u) u_x + b(x,y,u) u_y = c(x,y,u) .$$


El mismo analsis que nos llevo a construir un vector $v=(a,b,c)$
paralelo aun vector tangente $t=(dx/d\sigma, dy/d \sigma, du/d \sigma)$
nos llevo a las caracteristicas.


$$ \frac{dx}{d \sigma } = \lambda a(x,y,u)  $$
$$ \frac{dy}{d \sigma } = \lambda b(x,y,u)  $$
$$ \frac{du }{d \sigma } = \lambda c(x,y,u)  $$

De forma simple, nemotecnica.

$$ \frac{dx}{a} = \frac{dy}{b} = \frac{du}{c} .$$

DE estas ecuaciones podemos dividr la segunda por la primera
y la tercera por la primera


\begin{eqnarray}
\frac{dy}{dx} = \frac{b(x,y,u)}{a(x,y,u)} \quad , \quad
\frac{du}{dx} = \frac{c(x,y,u)}{a(x,y,u)} .
\end{eqnarray}

### Especificacion de datos.
Copiando de las ecuaciones lineales. Se piensa en curvas $\tau=\tau(\sigma)$,
La condicion inicial (IVP:Cauchy problem)

$$ x=x(0, \tau) \quad , \quad y = y(0, \tau) \quad , \quad u=u(0, \tau).$$

Resuelva con estas condiciones iniciales y halle


$$ x=x(\sigma, \tau) \quad , \quad y = y(\sigma, \tau) \quad , \quad u=u(\sigma, \tau).$$

Aca $u=u(\sigma, \tau)$ pero necesitamos $u(x,y)$ debemos resolver el sistema

\begin{eqnarray}
x &=& x(\sigma, \tau) \\
y &=& y(\sigma, \tau) \\
\end{eqnarray}
Para que este sistema tenga solucion necesitamos

$$ J = \begin{vmatrix} x_{\sigma} & x_{\tau} \\ y_{\sigma} & y_{\tau} \end{vmatrix} \ne 0 .$$

Reucuerden la condicion de no paralelismo.

Tenemos dos vectores. El vector de direcciones tangentes al plano $[a,b,c]$
y el tangente $\bf{t}=(dx/d\tau, dy/d \tau, du/d \tau)$
Si son paralelos los datos se especificaron a lo largo de la caracteristica.
El producto $\times$ no puiede ser cero. $\bf{t} \times \bf{v} \ne 0$.
Mas concretamente


$$\begin{vmatrix} \mathrm{e}_1 & \mathrm{e}_2 & \mathrm{e}_3 \\
\frac{dx}{d \tau} & \frac{dy}{d \tau} & \frac{du}{d \tau} \\ a & b & c \end{vmatrix}
=\left [  c \frac{dy}{d \tau} - b \frac{du}{d \tau} , a \frac{du}{d \tau} - c \frac{dx}{d \tau}, b \frac{dx}{d \tau} - a \frac{dy}{d \tau} \right ].$$

[Wikipedia](https://en.wikipedia.org/wiki/Envelope_(mathematics)) tiene un demo de esta situacion.


El concepto de *envlope of a family of curves curve* (envolvente de una familia de curvas) es el de que cada curva de la famila es **tangente** a la envolvente en un punto.  Matematicamente esto se traduce en el siguiente par de ecuaciones.
La familia de curvas debe parametrizarse con un parametro $\alpha$. De forma
que $y=f(x;\alpha)$ de forma explicita. $\alpha$ es la etiqueta de cada curva. Por ejemplo la pendiente en la grafica de arriba.

Las ecuaciones de la envolvente son

\begin{eqnarray}
\begin{cases}
 y = f(x ; \alpha) \\
 \frac{\partial y}{\partial \alpha} = 0
\end{cases}
\end{eqnarray}

La primera condicion es obvia pero la segunda condicion no es clara.


Vamos a generalizar mas el problem. Pensemos en la familia de forma **implicita**. Es decir piense en una superficie

$$F(x,y, \alpha)=0 .$$
En la forma **explicita** que tenemos arriba $F(x, y \alpha)=f(x ; \alpha)- y \equiv 0 $.

Las condiciones que debemos probar son

\begin{eqnarray}
\begin{cases}
 F(x,y, \alpha ) = 0 \\
 \frac{\partial F}{\partial \alpha} = 0
\end{cases}
\end{eqnarray}

Para hallar la envolmente, demos escribir la **familia** de curvas parametrizada por $\alpha$, **eliminar** $\alpha$ por medio de las dos ecuaciones anterior y escribir $y=y(x)$ como una curva explicita.

Para probar la segunda condicion necesitamos una figura.

Asumimos $F(x,y, \alpha_0)$. Nos enfocamos en la curva $\alpha_0$.

Tomando derivada todal de $F$ con respecto a $x$ y usando la regla de la cadena.
en la **envolente**,


$$ \frac{dF}{dx} = \frac{\partial F}{\partial x} \frac{\partial x}{\partial x} + \frac{\partial F}{\partial y} \frac{d y_{envolvente}}{dx} +  \left . \frac{\partial F}{\partial \alpha} \frac{d \alpha}{d x} \right |_{\alpha = \alpha_0} =0.$$

Ahora aplicamos la derivada pero en la curva $\alpha_0$.


$$ \frac{dF}{dx} = \frac{\partial F}{\partial x} \frac{\partial x}{\partial x} + \frac{\partial F}{\partial y} \frac{d y_{curve}}{d x} + \left . \frac{\partial F}{\partial \alpha} \frac{d \alpha}{d x} \right |_{\alpha = \alpha_0} = 0.$$

Cancelando terminos en la dos ecuaciones anteriores

Recuerden que $\alpha$ es constante para la curva verde y por eso el ultimo termino antes de la igualdad es 0


$$ \frac{\partial F}{\partial y} \frac{ d y_{envelope}}{dx} + \frac{\partial F}{\partial \alpha} \frac{d \alpha}{d x} = \frac{\partial F}{\partial y} \frac{d y_{curve}}{dx}  .$$

Ahora, la condicion de ser **envelop** es que las dos tangentes (en la verde y en la negra) son iguales


$$ \frac{d y_{envolmente}}{dx} = \frac{d y_{curva}}{dx} .$$
de forma que

$$ \frac{\partial F}{\partial \alpha} \frac{d \alpha}{d x } = 0.\ $$

Yo digo que por el teorema de la fucion inversa $d \alpha dx \ne 0$
para poder inverter $F(x,y,\alpha)$ pof $y=y(x)$.
