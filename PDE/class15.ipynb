## Funcion de Green para la ecuacion de onda unidimensional
Queremos resolver
$$u_{tt} - c^2 u_{xx} = F(x,t) $$
con condiciones iniciales (cero)

$$u(x,0) = 0 \quad , \quad u_t(x,0) = 0 $$
y de frontera

$$u(\pm \infty, t)= 0 $$

La funcion de Green satisface la siguiente ecuacion

$$G_{tt} - c^2 G_{xx} = \delta(x - \xi) \delta(t-\tau) $$
Tomamos la transfoda de Fourier en ambos terminos donde

$$\hat{G}(k, t, \xi, \tau) = \frac{1}{2 \pi} \int_{-\infty}^\infty G(x, t ; \xi, \tau) \mathrm{e}^{-\mathrm{i} k x} dx $$

Para obtener

$$\hat{G}_{tt} +  c^2 k^2 \hat{G} = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} \delta(t - \tau)  \tag{4} $$

Resolvememos la homogenea  y despues agregamos la fuente.

1. Solucion de la homogenea
$$\hat{G}_{tt} +  c^2 k^2 \hat{G} =  0 .$$
La resolvermos en terminos de senos y cosenos

$$\hat{G}(k, t, \xi, \tau) = A(k, \xi, \tau) \cos( ck(t - \tau) + B(k, \xi, \tau) \sin (ck(t - \tau). \tag{5} $$

2. Imponenmos condiciones iniciales.
Integramos la ecuacion (4) al rededor de un intervalo pequenho $[\tau-\epsilon, \tau + \epsilon]$

$$\int_{\tau - \epsilon}^{\tau + \epsilon} (\hat{G}_{tt} + c^2 k^2 \hat{G})
dt = \int_{\tau - \epsilon}^{\tau + \epsilon} \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i k \xi}} \delta(t - \tau) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}$$
=

De la mismo forma que vimos en la ecuacion de calor el segundo termino de la primera integral integra a 0 por ser continuo y por causalidad.
La integral queda


$$\hat{G}_t(k, \tau+) - \hat{G}_t(k, \tau^-) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}$$
Por las condiciones iniciales y causalidad eliminamos el segundo termino por que esta antes de la "explosion". Queda que exactamente en $t=\tau$

$$\hat{G}_t(k, \tau, \xi) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i}k \xi} \tag{5}$$

Aplicamos las condiciones iniciales para hallar $A,B$

Por la condicioal $0$ en $t = \tau$ y $\cos 0 =1$, $ \sin 0 = 0$ entonces
tenemos que $A=0$ de forma que

$$\hat{G}(k, t=\tau, \xi) = B(k, \xi, \tau) \sin(ck(t - \tau) $$
Todavia  necesitamos hallar $B$. Entonces la segunda condicion inicial, derivada en tiempo = 0

$$\hat{G}_t = B ck \cos (ck(t - \tau)) $$
si $t = \tau$
De la ecuacion (5)
$$ B = \frac{1}{2 \pi c k} \mathrm{e}^{-\mathrm{i} k \xi}  $$

La solucion en el dominio del numero de onda $k$ es:

$$\hat{G}(k, t ; \xi, \tau) = \frac{1}{2 \pi c k} \mathrm{e}^{-\mathrm{i} k \xi} \sin ( ck ( t - \tau)
 $$


Debemos sacar la transformada inversa para saber en espacio como es la funcion de Green

$$G(x,t; \xi, \tau ) = \frac{1}{2 \pi c} \int_{-\infty}^{\infty}
\frac{\sin( ck( t - \tau)}{k}  \mathrm{e}^{\mathrm{i}k (x - \xi) } dk$$

Esta integral no es simple de calcular. La calculo en el Apendice D.
El resultado es:


$$G(x, t ; \xi , \tau) = \frac12 \left (  H[ x- [\xi - c(t - \tau)] - H[x - [ \xi + c(t - \tau)]]
    \right )  $$

Ahora, teniendo en cuenta causalidad debemos multiplicar por $H(t - \tau)$.
Queda ento

$$G(x, t ; \xi , \tau) = \frac12 H(t - \tau ) \left (  H[ x- [\xi - c(t - \tau)] - H[x - [ \xi + c(t - \tau)]]
    \right )  \tag{6} $$

Si en parcular la fuente se encuentra en $\xi=\tau=0$. La formula se reduce a


$$G(x,t; 0, 0 ) = \frac12 H(t) [ H(x - ct) - H(x + ct)] $$
Esta es una version de la formula de deAlembert, y la interpretacion de (6) es la misma solo que fuente esta en otro lado $\xi$ en tiempo $\tau$.

## La funcion de Green para el operador Laplaciano.
La ecuacion a resolver es (ecuacion de Poisson):


$$\nabla^2 u(x,y) = f(x,y) $$

Para encontrar la funcion de Green debemos resolver


$$\nabla^2 G(x,y, \xi, \eta) = \delta(x - \xi) \delta(y-\eta) $$

y
$$u(x,y) = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} G(x,y, \xi, \eta) f(\xi, \eta) d \xi d \eta  $$

Para resolver este problema buscamos simetricas. Pensamos en coordenadas cilindricas (2D) $r=\sqrt{x^2 + y^2}$ Entonces, como queda el Laplaciano en coordenadas cilindriacas. (ejercicio de calculo en varias variables)

$$\frac{\partial r}{\partial x}= \frac{x}{r} \quad , \quad \frac{\partial r}{\partial y} = \frac{y}{r} $$
Ahora con $r \ne 0$. Then usando la regla de la cadena


$$ \frac{\partial u}{\partial r}  =  u'(r) \frac{x}{r} \quad , \quad
\frac{\partial u}{\partial y} = u'(r) \frac{y}{r} $$

Ahora vamos para las segundas derivadas.

$$\frac{\partial x/r}{\partial x} = \frac{r - x (x/r)}{r^2}= \frac1r - \frac{x^2}{r^3} $$

$$\frac{\partial^2 u}{\partial x^2} = u''(r) \frac{x^2}{r^2}+ u'(r)
\left (  \frac{1}{r} - \frac{x^2}{r^3} \right ) \tag{6} $$

Igualmente

$$\frac{\partial^2 u}{\partial y^2} = u''(r) \frac{y^2}{r^2}+ u'(r)
\left (  \frac{1}{r} - \frac{y^2}{r^3} \right ) \tag{7} $$

Sumando (6) y (7)

$$\nabla^2 u(r) = u''(r) \frac{x^2 + y^2}{r^2} + 2 \frac{u'(r)}{r} -u'(r) \frac{x^2 + y^2}{r^3} = u''(r)
 + \frac{u'(r)}{r}  $$

Simplificamos la ecuacion anterior a

$$u''(r) + \frac{u'(r)}{r} = \frac{1}{r} \frac{d}{dr} \left ( r \frac{du}{dr} \right )  $$

Asumiendo la fuente en $r_0$

$$\nabla^2 G(r, r_0) = \frac{1}{r - r_0} \frac{d}{dr} \left ( (r - r_0) \frac{du}{dr} \right ) =
\delta(r - r_0).  \tag{8} $$

Resolvamos este sistema (homogeneo)

\begin{eqnarray}
\frac{1}{r - r_0} \frac{d}{dr} \left ( (r - r_0) \frac{du}{dr} \right )  = 0  \\
\frac{d}{dr} \left ( (r - r_0) \frac{du}{dr} \right )  = 0  \\
 (r - r_0) \frac{du}{dr}  = C_1  \\
\frac{du}{dr}  = \frac{C_1}{r - r_0}  \\
G(r, r_0) = C_1 \ln (r - r_0) + C_2.
\end{eqnarray}

Continuamos en 15 minutos.

Para hallar las constantes $C_1, C_2$ necesitamos trabajar en el sistema no homogeneo mediante integracion.






Integramos la eucacion (8) en una esfera de radio $\epsilon$ al rededor de $r_0$.

$$\int_{B_\epsilon(r_0)} \nabla^2 G(r, r_0) d A = 1 $$
Usando el teorema de Guass en calculode varias variables
tenemos

$$\int_{\partial B_\epsilon(r_0)}  \nabla G \cdot n ds = 1 \tag{7}$$

Sabemos que

$$\nabla G(r, r_0) = \frac{C_1}{r - r_0} \hat{r}  $$
Reemplazamos esto en  (7)

$$\int_{\partial B_\epsilon(r_0)} \frac{C_1 ds}{\epsilon} = C_1 \frac{2 \pi \epsilon}{\epsilon} = 2 \pi C_1  $$


De forma que de (7) y la ultima ecuacion $2 \pi C_1 = 1$,

$$ C_1 = \frac{1}{2 \pi}$$.

Las condiciones de frontera al infinito dicen que $u_{r \infty}(r) = 0$
entonces $C_2 = 0$ y

$$G(r, r_0) = \frac{1}{2 \pi} \ln | r  - r_0 |. $$


Noten que no se uso transformada de Fourier.

## Separacion de variables en la ecuacion de onda en 2D en espacio y tiempo.
$(x,y) \in \in [0, L_x] \times [0, L_y]  $
Tenemos

$$u_{tt} - c^2(u_{xx} + u_{yy}) = 0 \quad \text{PDE}  \tag{8} $$
$$u(0, y, t) = u(L_x, y, t) = u(x,0,t) = u(x, L_y, 0) = 0  \quad , \text{Direchlet boundary conditions}$$
$$u(x,y,0) = \varphi(x,y) \quad , \quad u_t(x,y,0)=\psi(x,y)=0 \quad , \text{condiciones iniciales}$$

La idea es la misma. Asuma (separacion de variables)


$$u(x,y,t) = X(x) Y(y) T(t) $$
Al insertar esto en la ecuacion de onda (8) obtenemos

$$X(x) Y(y) T''(t) = c^2 ( X''(x) Y(y) T(t) + X(x) Y''(y) T(t)) $$
Dividiento dodo por $X(x) Y(t) T(t)$, encontramos

$$ \frac{T''(t)}{c^2 T(t)} = \frac{X''(x)}{X(x)} + \frac{Y''(y)}{y(y)}  $$.
Como cada lado depende de variables distintas cada lado es constante.
Como el segundo lado es una suma de funciones de variables distintas y esta suma es constante, cada termino es constate.
Tenemos entonces

$$\frac{T''(t)}{T(t)} = -c^2 ( \lambda+ \mu) = -\omega^2 $$
Renombramos

* $\lambda = k_x^2$, $\mu= k_y^2$,
* Encontramos

$$ k_x^2 + k_y^2 = \left ( \frac{\omega}{c} \right )^2 $$
Esta es la relacion de **dispersion** de la ecuacion de onda acustica.

Encontramos tres ecuaciones

$$T''(t) + \omega^2 c^2 T(t) = 0  $$
$$X''(x) + \lambda c^2 X(x) = 0  $$
$$Y''(Y) + \mu c^2 Y(y) = 0  $$
Las condiciones de borto $u(x,y,t) =0$ ,apara $x,y$ en los bordes ($0, L_x, L_y)$
Hablemos de la solucion en $X,Y$
$$X(0) = X(L_x) = 0  \quad ,  Y(0) = Y(L_y) = 0 $$
Sabemos que las soluciones son seno.  de la seccion 4.2 de las notas de clase

$$X(x) = C_1 \sin \left (  \frac{n \pi x}{L_x} \right ) \quad , \quad Y(y) = C_2 \sin \left ( \frac{n \pi y}{L_y} \right ) $$
Producto

$$X_n(x) Y_m(y) = C_{nx} C_{ny} \sin \left (  \frac{n \pi x}{L_x} \right ) \sin  \left (  \frac{n \pi y}{L_y} \right ) $$

Resolvamos para el dominio $T$
$$T''(t) + c^2 \omega^2 T(t) = 0 $$
La solucion es de la forma

$$T(t) = A \cos ( \omega t)  + B \sin ( \omega t) $$
De la condicion $T'(0)=0$, tenemos

$$ T'(0) = \omega B \cos \omega 0 = 0  \implies B = 0$$
En este caso tenemos

$$T(t) = A \cos ( \omega t) $$
generalizando a todos las autofunciones

$$T_n(t) = A_n \cos \omega_n t)$$

Vamos a usar superposicion y llamamos


$$w_{mn} = c \sqrt{ \left (  \frac{m \pi}{L_x} \right )^2 + \left ( \frac{n \pi y}{L_y} \right )^2 }{} $$
Entonces

$$u(x,y,t) = \sum_{m=1}^\infty \sum_{n=1}^\infty a_{mn} \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left ( \frac{n \pi y}{L_y}  \right ) \cos ( \omega_{mn} t) \tag{9} $$

Esta es la solucion, pero toca hallar $a_{mn}$.
Para esto aplicamos las condiciones iniciales

$$\varphi(x,y) = u(x,y,0) =
 \sum_{m=1}^\infty \sum_{n=1}^\infty a_{mn} \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left ( \frac{n \pi y}{L_y}  \right ).
 $$
 Probamos que las autofunciones (de dos indices)

 $$\varphi_{mn}(x,y) = \sin \left ( \frac{m \pi x}{L_x}  \right )
\sin \left (  \frac{n \pi y}{L_y}  \right )   $$
son ortogonales.

\begin{eqnarray}
\langle \varphi_{mn}(x,y), \varphi_{pq}(x,y) \rangle
&=& \left \langle \sin \left ( \frac{m \pi x}{L_x}  \right )
\sin \left (  \frac{n \pi y}{L_y}  \right ) , \sin \left ( \frac{p \pi x}{L_x}  \right )
\sin \left (  \frac{q \pi y}{L_y}  \right )   \right \rangle \\
&=&
\int_0^{L_x} \int_0^{L_y} \sin \left ( \frac{m \pi x}{L_x}  \right )
\sin \left (  \frac{n \pi y}{L_y}  \right ) \sin \left ( \frac{p \pi x}{L_x}  \right )
\sin \left (  \frac{q \pi y}{L_y}  \right )  dy \\
&=& \int_0^{L_x} \sin \left (  \frac{m \pi x}{L_x} \right )
\sin \left (  \frac{p \pi x}{L_x} \right ) dx \int_0^{L_y} \sin \left (  \frac{n \pi y}{L_y} \right )
\sin \left (  \frac{p \pi y}{L_y} \right ) dy \\
&=&  \frac{L_x}{2} \delta_{np} \frac{L_y}{2} \delta_{nq}.
\end{eqnarray}

Para hallar $a_[mn}$ proyectamos la condicione inicial $\varphi(x,y)$ sobre estas autofunciones

$$\langle \varphi(x,y) ,  \varphi_{pq}(x,y) \rangle = \sum_{m=1}^{\infty} \sum_{n=1}^\infty a_{mn} \langle \varphi_{mn}(x,y), \varphi_{pq}(x,y) \rangle=  \sum_{m=1}^{\infty} \sum_{n=1}^{\infty} a_{mn} \frac{L_x L_y}{4} \delta_{mp} \delta_{nq} = a_{pq} \frac{L_x L_y}{4}  $$







De donde y cambiando los nombres $m \to p$, $n \to q$o
$$a_{mn} = \frac{4}{L_x L_y} \int_0^{L_x} \int_0^{L_y} \varphi(x,y) \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left (  \frac{n \pi y}{L_y}  \right )  dy dx \tag{10} $$

La solucion es la (9) con coeficientes datos por la (10).

## Implementacion numerica

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# constantes
Lx, Ly = 1.0, 1.0
c=1.0
M, N = 15, 15 # numero de terminos de aproximacion
num_snapshots = 6
t_final=2.0
frames=100
dx = dy = 0.002

# create spatial grid
x = np.arange(0, Lx, dx)
y = np.arange(0, Ly, dx)
X, Y = np.meshgrid(x, y)

# condiciones iniciales. Gaussian
def phi(x,y):
    return np.exp(-400*((x-0.5)**2 + (y-0.5)**2))

# calcule los coeficientes de Fourier
def a_mn(m, n):
    A = 0
    for i in range(len(x)):
        for j in range(len(y)):
            A += phi(x[i], y[j])*np.sin(m*np.pi*x[i]/Lx)*np.sin(n*np.pi*y[j]/Ly) \
            *dx*dy
    return 4*A/(Lx*Ly)

# sume la solucin como series de m y n terminos
def u(x, y, t):
    U = np.zeros_like(x)
    for m in range(1, M+1):
        for n in range(1, N+1):
            a_mn_val = a_mn(m, n)
            omega_mn = c*np.sqrt( (m*np.pi/Lx)**2 + (n*np.pi/Ly)**2)
            U += a_mn_val*np.sin(m*np.pi*x/Lx)*np.sin(n*np.pi*y/Ly)*np.cos(omega_mn*t)
    return U

# create figure and axis for plotting
fig, ax = plt.subplots(2,  3, figsize=(12,8))

# generate snapshots and plot them
times = np.linspace(0, t_final, num_snapshots)

# counter
i=0
for ax, t in zip(ax.ravel() , times):
    print("i=",i)
    Z = u(X, Y, t)
    ax.contourf(X, Y, Z, levels=300, cmap='viridis')
    ax.set_title(f't = {t:.2f}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    i += 1

plt.tight_layout()
plt.show()



```

```python
# animacion
fix, ax = plt.subplots()
contour = ax.contourf(X, Y, u(X, Y, 0), levels=300, cmap='viridis')

def animate(i):
    ax.clear()
    Z = u(X, Y,t)
    contour = ax.contourf(X, Y, Z, levels=300, cmap='viridis')
    ax.set_title(f't = {t:.2f}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    print("i=",i)
    return contour

animation = FuncAnimation(fix, animate, frames=np.linspace(0, t_final, frames), interval=50)
```
