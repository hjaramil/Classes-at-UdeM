# continuacion de solucion de la PDE no lineal.

En la clase anterior mostramos que la solucion del la
ecuacion PDE $F(x,y,u,p,q)=0$ con $p=u_x, q=u_y$, $u=u(x,y)$
nos llevo a proponer siguiente sistema de ODEs

\begin{eqnarray}
\begin{cases}
\frac{dx}{d \sigma} = F_p \\
\frac{dy}{d \sigma} = F_q \\
\frac{du}{d \sigma} = p F_p + q F_q \\
 \frac{dp}{d \sigma} = - F_x - p F_u  \\
 \frac{dq}{d \sigma} = - F_y - q F_u
\end{cases} \tag{1}
\end{eqnarray}



Regresamos a la solucion de la PDE
$$u_x^2 + u_y^2 = u .$$
$$p^2 + q^2 = u .$$
$$F(x,y,u, p, u )= p^2 + q^2 - u \equiv 0 .$$

**Ejemplo 1:**
Tenemos $F_x = F_y = 0'$, $ F_u=-1$.
De las ultimas dos del sistema

$$  \frac{dp}{d \sigma} = p  .$$
$$  \frac{dq}{d \sigma} = q  .$$
Estas son faciles de resolver

$$ p = k_1 \mathrm{e}^{\sigma} \quad , \quad q = k_2 \mathrm{e}^{\sigma} .$$

Volvemos a las tres primeras de (1)

de

$$\frac{dx}{d \sigma} = 2 p \implies \frac{dx}{d \sigma} = 2 k_1 \mathrm{e}^{\sigma} .$$

$$\frac{dy}{d \sigma} = 2 q \implies \frac{dy}{d \sigma} = 2 k_2 \mathrm{e}^{\sigma} .$$

$$\frac{du}{d \sigma} = 2 u \implies \frac{dy}{d \sigma} = 2 u. $$

$$\frac{dx}{d \sigma} = 2 k_1 \mathrm{e}^{\sigma} .$$

$$ \frac{dy}{d \sigma} = 2 k_2 \mathrm{e}^{\sigma} .$$

$$\frac{du}{d \sigma} = 2 u.$$

Integramos estas ecuaciones

$$ x = 2 k_1 \mathrm{e}^{\sigma} + k_3 .$$
$$ y = 2 k_2 \mathrm{e}^{\sigma} + k_4 .$$
$$ u = k_5 \mathrm{e}^{2 \sigma} $$

En resumen tenemos
$$ p = k_1 \mathrm{e}^\sigma \quad , \quad q = k_2 \mathrm{e}^{\sigma}
\quad , \quad u = k_5 \mathrm{e}^{2 \sigma} \quad , \quad x = 2 k_1 \mathrm{e}^\sigma + k_3 \quad , \quad y = 2 k_2 \mathrm{e}^\sigma + k_4 .$$

Este es un sistema algebraico de ecuaciones no lineales en las variables (incognitas) $x,y,u,p,u$.

El problema esta en que las condiciones iniciales solo nos dan informacion sobre $x_0, y_0, u_0$.

Que hacemos con $p=p_0=p(\sigma=0)$, $q=q_0=q(\sigma=0)$?

Debemos resolver este problema y aparecen las

## Characteristic Strips (bandas caracteristicas).
El objeto de estas es encontrar las condiciones iniciales para $p,q$.

Asumimos que la superficie es $u=u(x,y)=u(x(\sigma, \tau), y(\sigma,\tau)$ y que conocemos
$x_0=x_0(\tau)$,   $y_0=y_0(\tau)$  $u_0=u_0(\tau)$  (para $\sigma=0$)

Para amarrar el problema a $p,q$, comenzamos con el plano tangente.


$$u-u_0 = p_0 ( x - x) + q_0(y-y_0) \tag{5} .$$

Tenemos entonces dos ecuaciones

\begin{eqnarray}
F(x_0, y_0, u_0, p_o , q_o)=0  \tag{2} \\
u-u_0 = p_0 ( x - x) + q_0(y-y_0) .
\end{eqnarray}
A lo largo de la curva inicial, que es parametrizada con $\tau$, usando
la regla de la cadena

$$ \frac{du_0}{d \tau} = p_0 \frac{d x_0}{d \tau} + q_0 \frac{d y_0}{d \tau}
\tag{3}.$$

Para poder invertir las ecuaciones (2) y (3) por $p_0, q_0$,
se requiere

$$ J= \begin{vmatrix} \frac{d x_0}{d \tau} & \frac{d y_0}{d \tau} \\ \frac{d F_0}{d p} & \frac{d F_0}{d q} \end{vmatrix} \ne 0   $$

Las condiciones de arriba se llaman **generalized transersally conditions** (condicones generalizadas transversales).

Los planos determinados por la Ecuacion (5) se llaman **characteristic strips** (bandas caraceristicas).




Como comparamos este sistema con el sistema quasi-lineal?

En el sistema quasi-lineal

$$F(x,y,u,p,q) = a(x,y,u) p + b(x,y,u) q - c(x,y,u) .$$

De aca
$$F_p =a(x,y,u) \quad , \quad F_q=b(x,y,u)$$, y el Jacobiano se reduce a


$$ \frac{d x_0}{d \tau} b - \frac{d y_0}{d \tau} a .$$
Este es el mismo jacobiano que vimos en las quasi-lineales.

Para resolver el problema, por pasos tenemos

1 **Paso 1**: Dado el IVP

\begin{eqnarray}
\begin{cases}
F(x,y,u,p,q) = 0 \\
x_0 = x(0, \tau) \quad , \quad y_0=y(0, \tau) \quad , \quad u_0=u_0(0, \tau)
\end{cases}
\end{eqnarray}
Debemos construir el $(p_0, q_0)$ iniciales  a partir de las ecuaciones


$$F(x_o, y_o, u_o, p_o, q_o)= 0 .$$
$$  \frac{d u_0}{d \tau} = p_0 \frac{d x_0}{d \tau} + q_o \frac{dy_o}{d \tau}
\tag{7} .$$

Con estas dos ecuaciones hallamos $p_0, q_0$ y esto completa las **cinco** condiciones iniciales.

Ilustramos esto con ejemplo que dejamos iniciado **Ejemplo 1**.

**Ejemplo 3**: La ecuacion eikonal.

Se usa para hallar tiempos a lo largo de los rayos en un medio acustico.  Si por ejemplo se puede medir $T$ como el tiempo a lo largo de un rayo, el gradiente del tiempo $T=T(x,y)$ debido a una fuente $(x_0, y_0)$.

[este link](https://www.weizmann.ac.il/complex/DOron/sites/complex.DOron/files/uploads/Geometrical%20optics%20notes.pdf) muestra una derivacion de la ecuacion eikonal para ondas electromagneticas.

$$\nabla T = \left [   \frac{\partial T}{\partial x}, \frac{\partial T}{\partial y}  \right ] .$$
$T$ es lo que hemos venido llamando $u$.
La ecuacion eikonal se deriva de asumir una onda plana en la ecuacion de onda,
asumiendo altas frecuencias. Esta derivacion (demostracion) esta por fuera de este curso.

$T$ mide del "slowness" (lentitud), la ecuacion eikonal dice que
$\| \nabla T \|^2 = \frac{1}{v^2(x,y)} $

En el lenguaje de nosotros tenems, asumiendo velocidad constante
$v(x,y)=c$

$$p^2 + q^2 = \frac{1}{v^2} .$$

Debemos postular 5 ODEs como se muestra en la Ecuacion (1)

Tenemos $F(x,y,u,p,q)= p^2 + q^2 - 1/v^2 = 0$
de forma

$$F_p = 2 p \quad , \quad F_q = 2 q \quad ,  \quad pF_p + q F_q = 2(p^2 + q^2)= \frac{2}{v^2} \quad F_x=F_y=F_u= 0$$



De la ecuacion (1)

$$\frac{dx}{d \sigma} = 2 p \quad , \quad \frac{d y}{d \sigma} = 2 q \quad , \quad \frac{du}{d \sigma} = \frac{2}{v^2}  \quad \frac{dp}{d \sigma} = \frac{d q}{d \sigma} = 0 .$$

Resolvemos por las condiciones iniciales $p=p_o$, $q=q_o$.
La tercera caracteristica
Las condiciones iniciales son

$$ x_0=x(0, \tau) \quad , \quad \sigma=0.$$
$$ y_0=x(0, \tau) \quad , \quad   \sigma=0.$$
$$ u_0=u(0, \tau) \quad , \quad   \sigma=0.$$

$$u = u_0 + \frac{2 \sigma}{v^2}  .$$
las ecuaciones para $x,y$

$$ x = x_0 + 2 p_0 \sigma \quad , \quad y = y_0 + 2 q_0 \sigma .$$

Debemos escribir $\sigma=\sigma(x,y)$, $\tau=\tau(x,y)$ en terminos de $x,y$,
de las ecuaciones:

\begin{eqnarray}
p_0^2 + q_0^2 &=& \frac{1}{v^2} \\
x &=& x_0 + 2 p_0 \sigma \\
y &=& y_0 + 2 q_0 \sigma \\
\end{eqnarray}

Si resuelve este sistema encontramos

\begin{eqnarray}
p_0 &=& \frac{x- x_0}{v \sqrt{ (x - x_0)^2 + (y - y_0)^2}} \\
q_0 &=& \frac{y- y_0}{v \sqrt{ (x - x_0)^2 + (y - y_0)^2}} \\
\sigma &=& \frac{v \sqrt{(x - x_0)^2 + (y - y_0)^2}}{2}
\end{eqnarray}

La solucion en terminos de $u$ es

$$ u = u_0 + \frac{\sqrt{(x - x_0)^2 + ( y - y_0)^2}}{v} .$$

En terminos de los tiempos a lo largo del rayo

$$ T = T_0 + \frac{\sqrt{(x - x_0)^2 + ( y - y_0)^2}}{v} .$$

Debemos intrepretar los resultados.

Las caracteristicas son

$$x(\sigma) = x_0 + 2 p_0 \sigma  \quad , \quad y(\sigma) = y_0 + 2 q_0 \sigma .$$
Estan de forma parametrica y las podemos escribir como



$$\frac{y-y_o}{x-x_o} = \frac{q_o}{p_o} = \text{constante} .$$
Esto es una recta de pendiente $q_0, p_0$.

Estas rectas son rayos. Y si $T=\text{constante}$ son "wavefronts" (frentes de onda).

# Ecuaciones diferenciales paciales de segundo orden (second order PDEs)

La ecuacion general parcial de segundo orden es de la forma

$$F(x_1, x_2, \cdots, x_n, u, u_{x_1}, u_{x_2} \cdots, u_{x_n}, u_{x_1 x_1}, u_{x_1 x_2}, \cdots u_{x_n x_n}) = 0 .$$


Otra forma
$$F(x_1, x_2, \cdots, x_n, u, u_{,1}, u_{,2} \cdots, u_{,n}, u_{,11},
u_{x,12 }, \cdots u_{,n n}) = 0 .$$
donde

$$u_{,ij} = \frac{\partial u}{\partial x_i \partial x_j} \quad , \quad i=1,2, \cdots, n \quad , \quad j=1,2, \cdots, n .$$

Por ejemplo

$$ L[u] = a u_{xx} + 2 b u_{xy} + c u_{yy} + d u_x + e u_y + fu = g .$$
La razon del 2 es:
* Asumimos que $u_{xy}=u_{yx}$ y se suman en $2$.
* El discriminante (lo vemos en un minuto) se "simplifica" si usamos el 2.

El discriminante de este sistema es

$D = (2 b)^2 - 4 ac = 4 (b^2 - ac) $
pero el 4 se ignora. O sea que
$D = b^2 - ac $

El discriminante sirve para **clasificar** las ecuaciones como sigue

1. **Elipticas**: $D < 0$, por ejemplo, la ecuacion de Laplace
$$ u_{xx} + u_{yy} = 0.$$

2. **Parabolicas**: $D=0$, por ejemplo la ecuacion de calor.
$u_t = \alpha u_{xx}

3. **Hiperbolicas**: $D>0$. Por ejemplo la ecuacion de onda

$$u_{tt} = \alpha u_{xx}$$

## Ecuaciones Hiperbolicas.
### Coeficientes constantes.
Asumimos

$$Lu = a u_{tt} + 2 b u_{tx} + c u_{xx} = 0 \quad , x \in \mathbb{R}, t> 0 .$$
$D>0$, $b^2 > ac \tag{11} $.
Pensamos en condiciones iniciales IVP:


\begin{eqnarray}
u(x,0) &=& \varphi(x) \\
u_t(x,0) &=& \psi(x)
\end{eqnarray}
Para facilitar el analisis, vamos  llamar

$$T = \frac{\partial }{\partial t}  \quad , \quad
X = \frac{\partial }{\partial x} .$$
Entonces la Ecuacion (11) se convierte en

$$ a T^2 + 2 b TX + c X^2 = 0. \tag{12} $$
Como esto ya es algebra, esta ecuacion tiene dos raices (piense en que $T$ es la incognita)

$$\lambda_{1,2} = \frac{-2 b X \pm \sqrt{4 b^2 X^2 - 4 ac X^2}}{2 a}
 =\frac{-b \pm \sqrt{b^2 - ac}}{a} X.$$

De esta forma podemos factorizar (12) como

$$
\left ( \frac{\partial }{\partial t} - \lambda_1 \frac{\partial}{\partial x} \right)
\left ( \frac{\partial }{\partial t} - \lambda_2 \frac{\partial}{\partial x} \right)  = 0.
$$

Esto da lugar dos ecuaciones diferenciales parciales

$$ \left ( \frac{\partial }{\partial t} - \lambda_1 \frac{\partial}{\partial x} \right)  = 0 $$

$$ \left ( \frac{\partial }{\partial t} - \lambda_2 \frac{\partial}{\partial x} \right)  = 0 $$


Usanso el metodo de las caracteristicas vemos que


$$\frac{dx}{dt} = \lambda_1 \quad , \quad \frac{dx}{dt} = \lambda_2 .$$
