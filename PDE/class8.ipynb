# Ecuacion de onda en cuerda semi-infinita

## Problema mixto.
$$u_{tt} = c^2 u_{xx} \quad , \quad x>0 \quad , t>0  $$
Condiciones iniciales
$$ u(x,0)= \varphi(x) \quad , \quad u_t(x,0) = \psi(x) \quad , x> 0 $$
De frontera
$$u(0, t)=0 \quad t > 0 $$
Este es el problema de Direchlet.

El metodo de d'Alembert establece que


$$u(x,t) = \frac{\varphi(x+ct) + \varphi(x-ct)}{2}  + \frac{1}{c} \int_{x-ct}^{x+ct} \varphi(s) ds.$$

Esta ecuacion es valida siemre que $x> ct$, pues $x-ct>0$, y obviamente $x+ct>0$
El problema esta en $x-ct$ que podria llegar a ser negativo.
Como arreglamos esto.

El principio de d'Alembert establece

$$u(x,t) = f(x-ct) + g(x+ct) $$
$x=0$


$$u(0,t) = f(-ct) + g(ct) $$
Lo idica que

$$g(ct) = - f(-ct) $$
Tenemos que $g$  esta definido en el lado positivo siempre que $f$ este en el lado negativo (y con signo contrario)

De las condiciones iniciales

$$u(x,0) = f(x) + g(x) = \varphi(x) \tag{1}) .$$
$$u_t(x,0) = - c f'(x) + c g(x) = \psi(x) $$
interamos la ultima ecuacion

$$-c f(x) + c g(x) = \int_0^x \psi(s) ds $$
multiplicando la (1) por $c$ y sumando a esta

$$2 c g(x) = c \varphi(x) + \int_0^x \psi(c) ds $$



De form que

$$ g(x) = \frac{\varphi(x)}{2} + \frac{1}{2c} \int_0^x \phi(s)$$
Para el caso      $-s < 0$, tenemos
que si $x- ct < 0$,

$$f(x-ct) = - g(ct - x) = - \frac{\varphi(ct-x)}{2} - \frac{1}{2c} \int_0^{ct -x} \varphi(s) ds = -\frac{\varphi(ct -x)}{2} + \frac{1}{2c} \int_{ct -x}^ 0$$



$$u(x,t) = f(x -ct) + g(x + ct) = \frac{\varphi(x+ ct) - \varphi(ct -x)}{2} + \frac{1}{2c} \int_{ct-x}^{x + ct} \psi(s) ds $$

Encontramos las dos ramas

\begin{eqnarray}
u(x,t) =
\begin{cases}
\frac{\varphi(x+ct) + \varphi(x-ct}{2} + \frac{1}{2c} \int_{x - ct}^{x + ct} \psi(s) ds \quad , \quad x > ct \\
\frac{\varphi(x + ct) - \varphi(ct-x)}{2} + \frac{1}{2c} \int_{ct -x}^{x + ct} \psi(s) ds \quad , \quad 0 < x < ct
\end{cases}
\end{eqnarray}

# Refleccion de ondas. Metodo de la extension par e impar (odd n even)
La reflecion par no la vemos en esta clase (se deja como tarea) pero se usa cuando las condiciones de bordo son sobre la derivada (Neumann). Para las condiciones direchlet se usa la extension impar.

La extender el problema (odd) es considerar todo el dominio $\mathbb{R}=(-\infty, \infty)$ por que alli yo se como operar (d'Alembert)

Como se estira el dominio?
Asumimos $\psi(x)=0$.

Condicion inicial extendida


\begin{eqnarray}
\hat{u}(x, 0) = \hat{\varphi(x) } =
\begin{cases}
\varphi(x)  \quad \text{si } x \ge 0  \\
-\varphi(-x) \quad , \quad \text{si } x < 0
\end{cases}
\end{eqnarray}
La solucion de d'Alembert es

$$\hat{u}(x,t) = \frac{\hat{\varphi}(x + ct ) + \hat{\varphi}(x - ct)}{2} .$$

**Ejemplo 1**:
Sea

\begin{eqnarray}
\varphi(x) =
\begin{cases}
0  \quad , \quad x < 3 \\
x-3  \quad , \quad 3 \le x < 4 \\
5 -x   \quad , \quad 4 \le x < 5 \\
0 \quad , \quad x \ge 5
\end{cases}
\end{eqnarray}

Dibujar la onda en tiempos $1, 2, 3, 3.5, 4, 4.5

```python
import numpy as np
import matplotlib.pyplot as plt

# define phi
def phi(x):
   return np.piecewise(x, [x < 3, (x >= 3) & (x < 4), (x >= 4) & (x <=5), x>5],
    [0, lambda x: x-3, lambda x: 5-x, 0])

# domain in x
x = np.linspace(0, 6, 1000)
y = phi(x)


# create plot
plt.figure(figsize=(8,4))
plt.plot(x, y, 'k', linewidth=1.5)

#plot dashed lines
plt.plot([0, 4], [1, 1], 'k--', linewidth=1.5)
plt.plot([4, 4], [0, 1], 'k--', linewidth=1.5)
plt.title('phi(x)')
plt.xlabel('x')
plt.ylabel('phi(x)')


# etiquetas
plt.text(4.1, 1.05, r'$\varphi(x)$', fontsize=16)
plt.text(5.3, -0.1, r'$x$', fontsize=16)

# limites
plt.xlim(0, 6)
plt.ylim(0, 1.5)



plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# reflexion
def phihat(x):
    return np.piecewise(x, [x < 0, x >= 0], [lambda x: -phi(-x), lambda x: phi(x)])

# define x values
x = np.linspace(-6, 6, 1000)
y = phihat(x)

# create plot
plt.figure(figsize=(8,4))
plt.plot(x, y, 'k', linewidth=2.0)

# labeles
plt.text(4.1, 1.05, r'$\hat{\varphi}(x)$', fontsize=16)
plt.text(5.3, -0.2, r'$x$', fontsize=16)

xticks = np.arange(-6, 6, 1)
yticks = np.arange(0, 1.5, 0.5)
plt.xticks(xticks)
plt.yticks(yticks)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# set limits
plt.xlim(-6.5, 6.5)
plt.ylim(-1.5, 1.5)




# draw arrows
plt.annotate('', xy=(6.5, 0), xytext=(-6,0), arrowprops=dict(facecolor='k', shrink=0.05, width=1, headwidth=8))
plt.annotate('', xy=(0, 1.5), xytext=(0,-1.5), arrowprops=dict(facecolor='k', shrink=0.05, width=1, headwidth=8))
plt.show()
```

```python
# definimos uhat
c=1
def uhat(x, t):
    return (phihat(x + c*t) + phihat(x - c*t))/2

# define x values
x = np.linspace(-6, 6, 1000)
y = uhat(x, 0)

# t values
t_values = [1, 2, 3, 3.5, 4, 4.5]

# create plots for each t in different frames
fig, axes = plt.subplots(len(t_values), 1, figsize=(10, 15), sharex=True)

for i, t in enumerate(t_values):
    y = uhat(x, t)
    axes[i].plot(x, y, 'k', linewidth=2.0)
    axes[i].set_title(f't = {t}')
    axes[i].set_ylim(-1.5, 1.5)
    axes[i].set_xlim(-6.5, 6.5)
    axes[i].grid(True, which='both', linestyle='--', linewidth=0.5)
    if i == len(t_values) - 1:
        axes[i].set_xlabel(r'$x$')
    axes[i].set_ylabel(r'$\hat{u}(x, t)$', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# animacion
import matplotlib.animation as animation

# define x values
x = np.linspace(-10, 10, 1000)
y = uhat(x, 0)

# create plot
fig, ax = plt.subplots(figsize=(8,4))
ax.set_xlim(-10, 10)
ax.set_ylim(-1.5, 1.5)
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$\hat{u}(x, t)$', fontsize=14)

line, = ax.plot(x, y, 'k', linewidth=2.0)

# initialize plot
def init():
    line.set_data([], [])
    return line,

# animation function
def animate(i):
    t = i * 0.1
    y = uhat(x, t)
    line.set_data(x, y)
    return line,


# create animation
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=100, blit=True)

ani.save('uhat.gif', writer='imagemagick')
```

# Cuerda finita. (siguiendo a Bleistein)
Resolvemos el IBVP: Initial Boundary Value Problem.

\begin{eqnarray}
\begin{cases}
u_{tt} - c^2 u_{xx}  = F(x,t) \quad , -\ell < x < \ell  \quad, c \text{ es constante}  \\
u(x,0) = \varphi(x) \quad , \quad u_t(x,0) = \psi(x) \quad \text{iniciales}\\
u(\pm \ell, t) = h_{\pm}(t) \quad , h_{\pm}(0)=\varphi(\pm \ell) \quad , \quad \
\frac{d h_{\pm}}{d t}(0) = \psi(\pm \ell)
\end{cases}
\end{eqnarray}



Nos basamos en la siguiente grafica

image.png

Queremos hallar el valor de la solucion en el punto $P$. Para esto vamos a integrar la ecuacion de onda en el paralelogramo $PABC$ usando el teorema de Green, como ya lo hicimos hace un par de clases. (alli fue en el triangulo verde, aca es  en el paralelogramo)

Habiamos dicho que

$$-\int_{PABC} F(x,t) dx dt = \oint_\Gamma u_t dx + c^2 u_x dt
= \int_{PA} + \int_{AB} + \int_{BC} + \int_{CP} u_t dx + c^2 u_x dt $$



Calculemos la primera integral. En el camino $PA$ la caracteristica es $x=x_0 + ct$, o sea $dx = c dt$, $dt=dx/c$.


$$\int_{PA} u_t dx + c^2 u_x dt = \int_{PA} c u_t dt + c u_x dx = c \int_{PA} u_t dt + u_x dx = c \int_{PA} du = c (u(A)-u(P)).  $$Calculemos

la segunda integral. En el camino $AB$ la caracteristica es $x=x_0 - ct$, o sea $dx = -c dt$, $dt=-dx/c$.



$$\int_{AB} = -c \int_{AB} du = c (u(A)-u(B)).  $$

la tercera integral. En el camino $BC$ la caracteristica es $x=x_0 + ct$, o sea $dx = c dt$, $dt=dx/c$.




$$\int_{BC} = c \int_{BC} du = c (u(C)-u(B)).  $$


La cuarta en $CP$, $x = x_0 - ct$, $dx=-cdt$, $dt=-dx/c$,
$$\int_{CP} = -c \int_{CP} du = c (u(C)-u(P)).  $$

Combinando las cuatro tenemos:

$$-\int_{PABC} F(x,t) dx dt = c( u(A)-u(P) + u(A)-u(B) + u(C)-u(B) + u(C)-u(P) =c( 2 u(A) - 2 u(P) - 2 u(B) + 2 u(C)  .$$



Despejando $u(P)$,

$$u(P) = u(A) + u(C) - u(B) + \frac{1}{2c} \iint \limits_{PABC} F(x,t) dx dt $$

A esta regla se le llama la **regla del paralelogramo** por que dice que el campo (suma) en dos vertices (contrarios) del paralelogramo es igual al campo
en los otros dos vertices salvo la fuente.
Es decir

$$u(p) + u(B) = u(A) + u(C) + \frac{1}{2c} \iint \limits_{PABC} F(x,t) dx dt .$$
Si la fuente es $F=0$ tenemos

$$u(p) + u(B) = u(A) + u(C)  $$
