Para entender el tema de separacion de variables necesitamos entender el Sturm-Liouville problem.

Definiciones basicas.

**Autavalor/Autofuncion* : Sea $L$ un operador lineal. Se dice que $\lambda \in \mathbb{C}$ es un **autovalor** (eigenvalue) y $u$ una autofucion (eigenfunction) si $Lu = \lambda u$, $u \ne0$. Si $u=0$ thenemos la
identidada $L0=0$. Esto no dice nada.

Sturm-Liouville descubrieron el siguiente operador

$$L u = \frac{1}{w(x)} \frac{d}{dx} \left (  p(x) \frac{du}{dx} \right )
+ q(x) u = \lambda u  \quad , \quad a < x < b $$
Esto reune una gama infinita de posibilidades para $w,p,q$ y dominio $D$.

Condiciones de frontera
$$\alpha_1 u(a) + \alpha_2 u'(a) = 0  \quad , \quad \beta_1u(b) +
\beta_2 u'(b) = 0 $$
No quereos $0=0$ deseamos $\alpha_1^2 + \beta_1^2 \ne 0$
o  $\beta_1^2 + \beta_2^2 \ne 0$
A estas se les llaman **Robin** boundary conditions.

Estas se puden reduce a condiciones que ya conocemos


$$u(a) = 0 = u(b) \quad, \quad \alpha_2=\beta_2=0 \quad , \quad \text{Direchlet}   .$$
$$u'(a) = 0 = u'(b) \quad, \quad \alpha_1=\beta_1=0   \quad , \quad \text{Neumann}.$$


Pensemos en el operador lineal
$$L u = \frac{1}{w(x)} \frac{d}{dx} \left (  p(x) \frac{du}{dx} \right )
+ q(x) u = \lambda u  \quad , \quad a < x < b $$
Es lineal por que $L(\alpha u + v) = \alpha L(u) + L(v)$. Ademas
$Lu=\lambda u$ definie auto-funciones y autovalores propios como veremos a continuacion.
A este operador lo llamamos el STO (Sturm-Liouville operator)
y al problema  lo llamamos el STP (Sturm-Liouville problem)

**Ejemplo 1.0**: Considere el siguiente problema
$$-u''(x) = \lambda u(x)  \quad 0 < x < \pi $$
$$u(0) =0   \quad , u(\pi) = 0 $$

Encontramos que

$$\text{ for } \lambda=1 \implies -u''(x) - \lambda^2 u(x) = 0 $$

Recordemos de los cursos de ODE. El polinimo caracteristio asociado

$$D^2 + \lambda^2 = 0 $$
Las raices del polinomio eson $\pm \mathrm{i}$.
Factorizamos

$$(D- \lambda \mathrm{i}) (D + \lambda \mathrm{i}) = 0 $$
De $D- \mathrm{i}$ = 0

$$ (D - \lambda \mathrm{i}) u = 0$$
$$ \frac{du}{dx} - \lambda \mathrm{i} u = 0$$
$$ \frac{du}{dx} = \lambda \mathrm{i} u = 0$$
$$ \frac{du}{u} = \lambda \mathrm{i} $$
$$ u=\mathrm{e}^{\mathrm{i} \lambda x } $$
Igualmente para el $-$
$$ u=\mathrm{e}^{-\mathrm{i} \lambda x } $$
Combinamos las dos anterioreso

$$u(x) = C_1 \mathrm{e}^{\mathrm{i} \lambda x } + C_1 \mathrm{e}^{\mathrm{i} \lambda x } $$
de $u(0)=0$

$$u(0) = C_1 + C_2 = 0 $$
O sea que $C_2 = -C_1$

Si $C_1 \ne 0$ debe ser que $\sin \lambda \pi =0$. De forma que



$$ \lambda = \lambda_n =n $$
son un numero infinito de autovalores de del operador $L$ con condiciones iniciales. Las auto-funciones (eigenfunction) son

$$\{ \sin n \pi x  \} $$

En espacios finitos $A \in \mathbb{R}^{n \times n}$ se define el sistema


$$ Ax = b$$
aca el problema

Para poder **generalizar** lo que aprendimos en algebra lineal al SLP, necestiamos definir algunos conceptos.

### Producto interno
Sean $f,g$ dos funciones en el espacio de interes ($L^2(a,b)$)
sea $w>0$ y sea $D$ el dominio donde se encuentran nuestras funciones
Definimos el **producto interno** con la ecuacion

$$\langle f , g \rangle_w = \int_D w(x) f(x) \overline{g(x)} dx  .$$
donde $\overline{g}$ es el **conjugado** de $g$.
$D$ pues ser cualquier intervalo incluyendo $(-\infty, \infty)$
El producto interno es bilineal (lineal en cada una de sus partes

$$\langle f + g , h \rangle = \langle f, h \rangle + \langle g, h \rangle \quad , \quad \text{and} $$
$$\langle f , g + h \rangle = \langle f, g \rangle + \langle f, h \rangle $$

### Normal de una funcion
Definimos la norma $\ell_2$ como
$$ \| f \| = \sqrt{ \langle f , f \rangle } $$
### Ortogonalidad
Dos funciones $f \ne g$ son ortogonales si
$$\langle f, g \rangle = 0 $$

**Ejemplo 2.0**:

$$\sin i x \sin jx = \frac{ \cos(i-j)x - \cos (i+j) x }{2} $$

De igual forma la segunda integral tambien da 0,

###  Operador adjunto
El operador adjunto es la extension de lo que es una matriz **traspuesta** o **traspuesta conjugada** si la matriz fuera compleja.
El ajunto (adjoint) de un operador diferencial $L$ es $L^*$ tal que

$$\langle Lu , v \rangle = \langle u , L^* v \rangle $$
para todo $u$ en el dominio de $L$ y $v$ en el dominio de $L^*$

**Ejemplo3.0**: Sea el problema STL

\begin{eqnarray}
Lu &=& \frac{du}{dx} \\
u(0)&=&2 u(1)
\end{eqnarray}
Encuentre el opeador adjunto de $L$, es decir $L^*$. El truco en este tipo de ejercicioes es **integration por partes**.

**Solucion:**
Por definicion
$$\langle u, v \rangle = \int_0^1 u(x) v(x) dx $$
Entones

\begin{eqnarray}
\langle Lu, v \rangle &=& \int_0^1 u'(x) v(x) dx \\
&=& u(1) v(1) - u(0) v(0) - \int_0^1 u(x) v'(x) dx \\
&=& u(1)(v(1) - 2v(0)) - \int_0^1 u(x) v'(x) dx \\
&=& u(1) ( v(1)- 2 v(0)) + \langle u, L^* v \rangle
\end{eqnarray}
donde
$$L^* v = - \frac{dv}{dx} $$
Es decir

$$\langle L, u, v \rangle = \langle u, L^*v \rangle + u(1) (v(1)- 2 v(0)] \tag{1} $$

Si queremos que $\langle Lu, ,v \rangle = \langle u, L^*v \rangle$ necesitamos
al condicion de frontera en $v$ omo $v(1)=2 v(0)$ entonces la ecuacion
se convierte en $\langle L u, v \rangle = \langle u , L^* v \rangle$.
Y el adjunto  de $L$ se consiguie como

Queremos que $J(u,v)=0$ para que el operador sea autoadjunto.





\begin{eqnarray}
L^*v = - \frac{dv}{dx} \\
\text{con condiciones de borde} \quad v(1) = 2 v(0).
\end{eqnarray}

De lo que vimos encontramos

$$\langle Lu, v \rangle = J(u,v) + \langle u, L^* v $$

**Ejemplo 3**: Enucuentre el adjunto de

\begin{eqnarray}
\begin{cases}
Lu = a_2(x) \frac{d^2u}{dx^2} + a_1(x) \frac{du}{dx} + a_0(x) u \\
\alpha_1 u(a) + \beta_1 \frac{du(a)}{dx}=0 \quad , \quad \alpha_2 u(b) + \beta_2 \frac{du(b)}{dx} = 0
\end{cases}
\end{eqnarray}

Usamos $u'(x)=du/dx$, $u''=d^2u/dx^2$, y formamos $\langle Lu, v \rangle $ como sigue


\begin{eqnarray} \langle Lu , v \rangle &=& \int_a^b [a_2(x) u''(x) + a_1(x) u'(x) + a_0(x) u(x)] v(x) dx  \\
&=& \int_a^b [a_2(x) v(x)] u''(x) + [a_1(x) v(x)] u'(x) + [a_0(x) v(x) u(x)] dx
\end{eqnarray}

Vamos por partes

\begin{eqnarray} \int_a^b [a_2(x) v(x)] u''(x) dx &=&
\left .  [a_2(x) v(x)] u'(x) \right |_a^b  -\int_a^b u'(x)
[ a_2(x) v(x)]' dx \\
&=&  \left .  [a_2(x) v(x)] u'(x) \right |_a^b  \left . -[a_2(x) v(x)]' u(x) \right |_a^b + \int_a^b [a_2(x) v(x)]'' u(x) dx    \\
\end{eqnarray}
La segunda integral es

$$\int_a^b [a_x(x) v(x)] u'(x) dx = \left . [a_1(x) v(x) u(x) ]  \right |_a^b - \int_a^b [a_1(x) v(x)]' u(x) dx$$


Combinamos todos los terminos.
Todos los terminos de frontera son

$$J(u,v) =  \left .  [a_2(x) v(x)] u'(x) \right |_a^b  \left . -[a_2(x) v(x)]' u(x) \right |_a^b +  \left . [a_1(x) v(x) u(x) ]  \right |_a^b $$

Escribimos
$$\langle Lu , v \rangle = J(u,v) + \int_a^b [ a_2(x) v(x)]'' - [a_1(x) v(x)]' + a_0(x)] u(x) dx $$



Is obligamos $J(u,v)=0$ tenemos


$$\langle Lu, v \rangle = \langle u, L^* v \rangle $$
donde

\begin{eqnarray}
\begin{cases}
L^*v = \frac{d^2}{dx^2} [ a_2(x) v(x)]  - \frac{d}{dx} [ a_1(x) v(x)] + a_0(x) v(x) \\
J(u,v)= 0.
\end{cases}
\end{eqnarray}
Haciendo $J(u,v)=0$ obligamos a que



$\langle Lu, v \rangle = \langle u , L^*v \rangle$
al opeador $L^*$ sin cosiderar los terminos de bordo se le llama **ajunto formal** y con las condiciones de bordo $J(u,v)=0$ se llama **adjunto completo** (full adjoint).




### Operador autoajdunto.

Al operador $L$ se le llama a**autoadjunto** si $L=L^*$.


Como tarea le dejamos al estudiante que pruebe


Que el Sturm-Liouville operator es autoadjunto con

$$J(u,v) = \left . p(x) \left (  \frac{du}{dx} v(x) - \frac{dv}{dx} u(x)  \right ) \right |_a^b .$$


**Teorema**:
Si $L$ es autoajdunto sus valores propios son reales y sus funciones propias (eivenfunctions) son ortogonales.
Ausma $u_j$ autofunciones con autovalores $\lambda_j$
Sin perdida de generalidad podemos asumir que $\| u \|_j=1$.

$$ \langle L u_j, u_j \rangle = \langle u_j , L u_j \rangle = \langle u_j, \lambda u_j \rangle = \overline{\lambda_j} \langle u_j, u_j \rangle = \overline{\lambda_j} $$
Ahora bine
$$\langle L u_j, u_j \rangle = \langle \lambda_j u_j, u_j \rangle =\lambda \langle u_j, u_j \rangle = \lambda $$

De forma que $\overline{\lambda} = \lambda$, o sea que $\lambda \in \mathbb{R}$o

Para la segunda parte considere dos functions $u,v$ correspondientes con
autovalores distintos $\lambda_1, \lambda_2$

$$\langle Lu, v \rangle = \langle u , L v \rangle = \langle u, \lambda_2 v \rangle = \lambda_2 \langle u, v \rangle \tag{3} $$

$$\langle Lu, v \rangle = \langle \lambda_1 u , v \rangle = \lambda_1 \langle u, v \rangle   \tag{4} $$

De (3) - (4)
$(\lambda_2 - \lambda_1) \langle u, v \rangle = 0 $
esto quiere decir que $\langle u, v \rangle =0$ (pues $\lambda_2 - \lambda_1 \ne 0$

Definamos bases completas.
Para entender la ventaja de funciones ortogonales, bajemonos a $\mathbb{R}^2$,
piense en un vector $v = \binom{a}{b}$ y las bases ortogonales

$$\hat{\mathrm{e}}_1 = \binom{1}{0}  
 \quad \hat{\mathrm{e}}_2 = \binom{0}{1}  $$

 Entonces

 $$\binom{a}{b} = a \hat{\mathrm{e}_1}  + b \hat{\mathrm{e}}_2 =
 \langle v, \hat{\mathrm{e}}_1 \rangle \hat{\mathrm{e}}_1 +
 \langle v , \hat{\mathrm{e}}_2 \rangle \hat{\mathrm{e}}_2$$


 En general $v$ vive en un espacio infinito dimensional y lo podriamos
 escribir como

 $$v = \sum_{n=1}^{\infty} c_i \varphi_i \tag{5} $$
 donde $\varphi_i$ son auto-funciones (eigenfuctions). Diagmos que tengan norma 1 por el momento.

 Miren como extraemos los coeficientes $c_i$.

 $$  \langle v , \varphi_i \rangle = \sum_{j=1}^\infty c_j \langle \varphi_j, \varphi_i \rangle  =  \sum_{j=1}^\infty c_j \delta_{ij} = c_i \tag{6}$$

 A la serie infinita (5) se le llama **Serie de Fourier** y a los
 coficientes $c_i = \langle v, \varphi_i \rangle $ se les llama **coeficientes de Fourier**.
