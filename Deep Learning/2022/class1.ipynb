## Fundamentos matematicos

### Tensores
Los tensores son entes matematicos que representan fenomenos fisicos invariantes bajo cualquier transformacion de coordenadas. Nada de esto vamos a ver en DL. En DL los tensores son objetos multi-indice (como matrices o mas alla?)


#### Motivacion:
Un ejemplo: Tome una imagen (una fotografia) tiene pixeles $nx \times ny$, pero la fotografia podria tener canales (RGB: Reed Green Blue) 3 canales o mas entonces 
$nx \times ny \times nc$. La fotografia podria pertenecer a un indice de fografias para escoger 
$ni \times nx \times ny \times nc$ (rango 4)


Los tensores (matematicamente hablando) son transformaciones multilineales. Explico:

Notacion: A las transformacion (lineales) las llamamos $T$,
los vectores $x,y$ (los vectores van a ser tensores).


Hacemos un analisis (bottom-up) de abajo hacia arriba.

##### Transformaciones lineales:

* $ x \in \mathbb{R}, y \in \mathbb{R} $:

$$ T x = y = a x$$
aca todos son escalares, $x,y,a$ (**tensores** de rango 0)

* $x \in \mathbb{R}^n , y \in \mathbb{R}^m$

$$ T x = y = a x $$

Sucede que $a \in \mathbb{R}^{m \times n}$. En este caso
$y$ es un **tensor** de rango 1. Lo notamos $y_i$, la $i$ componente del tensor $y$. Quien es $a$? (repaso de algebra lineal).
Pensemos que $T$ es una transformacion lineal $T(c_1 x_1 + c_2 x_2) = c_1 T(x_1) + c_2 T(x_2) $, $c_1, c_2 \in \mathbb{R}$, $x_1,  x_2$ son vectores (en este caso $x_1, x_2 \in \mathbb{R}^n$ .  Definamos los vectores base

 \begin{eqnarray*}
      e_1 = \begin{pmatrix}
        1 \\
        0 \\
        \vdots  \\
        \vdots \\
        0
      \end{pmatrix} 
      \quad , \quad 
      e_2 = \begin{pmatrix}
        0 \\
        1 \\
        \vdots  \\
        \vdots \\
        0
      \end{pmatrix}
      \quad , \quad 
      e_k = \begin{pmatrix}
        0 \\
        \vdots \\
        1  \\
        \vdots  \\
        0 
      \end{pmatrix}
      \quad ,  \quad \cdots  , \quad 
      e_n = \begin{pmatrix}
        0 \\
        \vdots  \\
        \vdots \\
        0 \\
        1
      \end{pmatrix} \; .
    \end{eqnarray*}
 Podemos escribir $x = \sum_{i=1}^n x_i e_i $, como $T$ es lineal

 $$y= Tx  = T \left ( \sum_{i=1}^n x_i e_i \right ) = \sum_{i=1}^n x_i T(e_i)  $$

 segun eso $y = a x$, donde $a$:

\begin{eqnarray}
      a = \begin{pmatrix}
        | & | & \vdots & \vdots & | \\
        T(e_1)   & T(e_2) & \vdots & \vdots & T(e_n)   \\
        | & | & \vdots & \vdots & | 
      \end{pmatrix} \, . 
\end{eqnarray}

$T(e_i) \in \mathbb{R}^m$.

Nos metemos ahora al algebra **multilineal**

* $x \in \mathbb{R}^{n \times m}, y \in \mathbb{R}^{p \times q} $

$$ Tx = y = ax  $$

Quien es $a$.
Veamos: Empecemos con un ejemplo $n=m=2$. Estos estan dados por

   \begin{eqnarray*}
      e_{11} = 
      \begin{pmatrix}
        1 & 0 \\
        0 & 0
      \end{pmatrix}
      \quad , \quad 
      e_{12} = 
      \begin{pmatrix}
        0 & 1 \\
        0 & 0
      \end{pmatrix}
      \quad , \quad 
      e_{21} = 
      \begin{pmatrix}
        0 & 0 \\
        1 & 0
      \end{pmatrix}
      \quad , \quad 
      e_{22} = 
      \begin{pmatrix}
        0 & 0 \\
        0 & 1
      \end{pmatrix} .
    \end{eqnarray*}






Los vectores base son en verdad matrices base. 

Definimos el **producto tensorial** de la siguiente forma

$$c_{ij} = (a \otimes b)_{ij} = a_i b_j  $$
$a,b$ son vectores en $\mathbb{R}^n$. 

Un ejemplo:

 \begin{eqnarray*}
      e_{11}  = e_1 \otimes e_1 = 
      \begin{pmatrix}
        1 \times 1 & 1 \times 0 \\
        0 \times 1 & 0 \times 0 
      \end{pmatrix}
      = \begin{pmatrix}
        1 & 0  \\
        0 & 0
      \end{pmatrix} \; . 
    \end{eqnarray*}
$e_1=(1,0)^T$ 

Podemos generalizar

El objeto $e_i \otimes e_j$ ($e_i, e_j \in \mathbb{R}^n$)
es una matriz:




Selection_004.png

Estas matrices son los "vectores" base en el espacio vectorial $\mathbb{R}^{n \times m}$.  Del algebra lineal

$$x = \sum_{i=1}^n \sum_{j=1}^m x_{ij} e_{ij}  $$

Aplicamos la transformacion $T$

$$ T(x) = \sum_{i=1}^n \sum_{j=1}^m x_{ij} T(e_{ij})$$

$T$ cae en el espacio $y=Tx=ax$, $T \in \mathbb{R}^{p \times q}$ .   El operator $T$ es una matriz

$$T_{st}(x) =   \sum_{i=1}^n \sum_{j=1}^m x_{ij} T_{st}(e_{ij})  \quad s=1,2, \cdots, p \quad , \quad t=1,2, \cdots ,q \quad , \quad  (1) $$

De forma el $a$ se define como

$$ a_{st}^{ij} = T_{st}(e_{ij})  $$

Ojo con los nombres:

Los indices $ij$ en la ecuacion (1) son **repetidos**
Los indices repetidos son aquellos sobre los cuales se suma y se llaman "dumies" (mudos) y los que no son repetidos (aparecen una sola vez) se llaman libres (free indices) 

Albert Eisten, en su teoria de la relatividad invento una notacion mas compacta. Se elimina el simbolo sumatoria $\Sigma$, la suma se asume sobre los indices repetidos (dummy, mudos)


$$T_{st}(x) =    x_{ij} T_{st}(e_{ij})  \quad s=1,2, \cdots, p \quad , \quad t=1,2, \cdots ,q \quad , \quad  (1) $$

Entonces

$$T_{st}(x) = x_{ij} T_{st}(e_{ij}) = a_{st}^{ij} x_{ij}  $$

Estamos listos para generalizar:

* $x \in \mathbb{R}^{n_1 \times n_2 \times \cdots \times n_j}$ y $y \in \mathbb{R}^{m_1 \times m_2 \times \cdots \times m_k}$

$$ T(x) = a y $$
quien es $a$?

Los vectores base $e_{p_1 p_2 \cdots p_j} = e_{p_1} \otimes e_{p_2} \otimes \cdots \otimes e_{p_j}$ donde $e_{p_i}$ es un vector con 1 en la posicion $p_i$ y 0 en las demas.

De forma que podemos escribir

$$T(x) = x_{p_1 p_2 \cdots p_j} T( e_{p_1 p_2 \cdots p_j}) $$


$$T_{i_1 i_2 \cdots i_k}(x) = x_{p_1 p_2 \cdots p_j} T_{i_2 i_2 \cdots i_k}( e_{p_1 p_2 \cdots p_j}) $$

El objeto $a$ lo representamos


$$ \left [ T(e_{p_1 p_2 \cdots p_j})\right ]_{q_1 q_2 \cdots q_k}  = a_{q_1 q_2 \cdots q_k}^{p_1 p_2 \cdots p_j} $$

queamos con la expresion

\begin{eqnarray*}
      T_{q_1 q_2 \cdots q_k} (x) = 
      a^{p_1 p_2 \cdots p_j}_{q_1 q_2 \cdots q_k} x_{p_1 p_2 \cdots p_j} \, .
\end{eqnarray*}

La tensor $ T_{q_1 q_2 \cdots q_k}$  se le llama tensor de
orden $k$ (covariante?)

Piense que $x \in \mathbb{R}^{n_1 \times n_2 \cdots \times n_j}$
piense que $x$ se puede como un tupla de vectores

$$x = (x_1, x_2, \cdots, x_j) $$

Para entender las dimensiones en DL. Supgona que se tiene 10 imagenes, cada imagen con $400 \times 600$ pixeles  y con 3 canales . Tenemos $10 \times 400 \times 600 \times 3 = 7200000$.
