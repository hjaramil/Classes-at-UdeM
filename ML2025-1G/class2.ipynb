Asumimos el teorema de Pitagoras

$$\| x \|^2 + \| y \|^2 = \| x + y \|^2 \tag{1} $$
Veamos como el teorema de Pitagoras implica que los vectores $x,y$ son ortogonales (es decir $\langle x, y \rangle=0$.

Veamos:

$$\| x + y \|^2 = \langle x+y, x+ y \rangle = \langle x, x \rangle + \langle x, y \rangle + \langle y, x \rangle + \langle y , y \rangle  = \| x \|^2 + 2 \langle x, y \rangle + \| y \|^2 \tag{2} $$

Igualando (1) con (2) vemos que

$$2 \langle x, y \rangle = 0 $$

$$ \langle x, y \rangle = 0 $$
Es decir $x,y$ son ortogonales.

**Definicion 2.1.4**(matriz identidad): La matriz $n \times n$ $I$ se llama la identidad si la entrada $ij$ esta definida por el **delta de Kronecker** $\delta_{ij}$.

$$\delta_{ij} = \begin{cases} 1 \quad \text{si  } i = j \\ 0 \quad  \text{de lo contrario, $i \ne j$}  \end{cases} $$

Es decir una matriz con diagonal unos y de resto ceros.

$$I = \begin{pmatrix}
1 & 0 & 0 & \cdots & 0 \\
0 & 1 & 0 & \cdots & \vdots  \\
0 & 0 & 1 & \cdots & \vdots  \\
\vdots & \vdots & \vdots & \ddots & 0 \\
0 & \cdots & \cdots & 0 & 1
\end{pmatrix} $$

Ojo con el **delta de Kronecker**. El delta de Kronecker **tumba** una sumatoria. Es decir, por ejemplo

$$\sum_{j=1}^n \delta_{ij} a_i = a_i $$
Verifique




$$\delta_{1i} a_1 + \delta_{2i} a_2 + \cdots \delta_{ii} a_i + \delta_{i+1 i} a_{i+1} + \cdots + \delta_{in} a_n= a_i$$

Si la suma fuera doble, se tumba una y queda sencilla.

$$\sum_{k=1}^n \sum_{j=1}^n \delta_{ij} a_{ik} = \sum_{k=1}^n a_{ik}. $$

**Teorema 2.1.8**: Si $AB=I$ y $A,B$ son matrices cuadradas. Entondces $BA=I$. O sea que la inversa conmuta.

Prueba: Asumimos que $AB=I$, y multiplicamos a $B$ for esta identidad, luego usamos la propiedad asociativa del producto

$$B = B (AB) = (BA) B.  \tag{3} $$
Como $\det(I)=1$, entonces $\det(AB)=\det(A) \det(B) = 1$, o sea que $B$ es invertible, pues $\det(B) \ne 0$ y podemos multiplicar la ecuacion, por $B^{-1}$ a la derecha

$$BB^{-1} = (BA)= B B^{-1} = BA.   $$
Es decir $I = BA$.

Entonces ya podemos decir que si $A$ es ortogonal entonces
$$A^* A = A A^* = I $$

## Autovalores/Autovectores (valores propios/vectores propios, Eigenvalues/Eigenvectors del Aleman Eigen (auto)

Comenzamos con la definicion

**Definicion 2.2.1**(autovalores/autovectores). Dada una matriz $A \in \mathbb{C}^{n \times n}$ y un vector  $u \in \mathbb{C}^n$ que satisfacen la ecuacion

$$A u  = \lambda u $$
con $\lambda \in \mathbb{C}$, decimos que $\lambda$ es un autovalor y que $u$ es un autovector de $A$.

[Recomiendo Wikipedia](https://upload.wikimedia.org/wikipedia/commons/0/06/Eigenvectors.gif)  animacion.

La pregunta siguiente es. Como se hallan los autovalores/vectores?  Veamos como resolver este problema.

$$A u  = \lambda u $$
$$A u  = \lambda I u $$
$$A u  - \lambda I u =0 $$
$$(A   - \lambda I) u =0 \tag{5} $$

Con esto observamos del algebra lineal (teoria de matrices y determinantes y sistemas lineales) que si $Bx=0$, entonces $\det B = 0$ si $x \ne 0$.
De forma

$$p(\lambda) = \det(A - \lambda I) = 0. $$
Para encontrar $\lambda$ buscamos las raices del polinomio caracteristico $p(\lambda)= \det(A - \lambda I)$.

Una vez hallado el autovalor $\lambda$, resolvemos la ecuacion (5) para encontrar $u$.
Note que hay infinitos $u$ por que el determinante de la matriz $A - \lambda I$ es 0, las filas son linealmente dependientes.

Vamos a ver ejemplos.

**Ejemplo 2.2.1**: Halle los autovalores/autovectores de la siguiente matriz

$$A = \begin{pmatrix} -5 & -3 \\ -6 & 2\end{pmatrix}  $$

Solucion: Comenzamos por hallar los autovalores $\lambda$.
El polinomio caracteristico es

\begin{eqnarray}
\det(A - \lambda I) &=& \det \begin{pmatrix}
-5 - \lambda & -3 \\ -6 & 2 -\lambda
\end{pmatrix} \\
&=& (-5 -\lambda)(2 - \lambda) - (-6)(-3) \\
&=& -10 + 3 \lambda + \lambda^2 - 18 \\
&=& \lambda^2 + 3 \lambda - 28 \\
&=& (\lambda + 7  )(\lambda - 4 ) = 0
\end{eqnarray}
Esto da 2 soluciones $\lambda=-7$, $\lambda=4$.

Como hallamos los autovectores?
1. $\lambda = -7$

Vamos a la ecuacion (5)
$$\left \{ \begin{pmatrix} -5 & - 3 \\ -6 & 2 \end{pmatrix}   - \begin{pmatrix} -7 & 0 \\ 0 & -7 \end{pmatrix} \right \} \binom{x}{y} = \binom{0}{0}  $$

$$ \begin{pmatrix} 2 & - 3 \\ -6 & 9 \end{pmatrix}    \binom{x}{y} = \binom{0}{0}  $$

Como el determinante es 0, una fila es multiplo de la otra (son linealmente dependientes). Tomemos solo una, por ejemplo, la primera

$$ 2 x - 3 y  = 0 $$
$$ y =  \frac{2x}{3}$$

Esto da como solucion

$$\binom{x}{2 x/3} $$
Por ejemplo si $x=3$ tenemos la solucion

$$ u = \binom{x}{y} = \binom{3}{2}$$

Si $u$ es autovector de $A$ entonces $\beta u$ tambien lo es pues
$$ A u = \lambda u$$
$$ A \beta u = \lambda \beta u$$
$$ A v  = \lambda v$$
con $v = \beta u$
y el autovalor es el mismo.
Por eso los autovectores se entregan normalizados.

2. $\lambda = 4$


$$\left \{ \begin{pmatrix} -5 & - 3 \\ -6 & 2 \end{pmatrix}   - \begin{pmatrix} 4 & 0 \\ 0 & 4 \end{pmatrix} \right \} \binom{x}{y} = \binom{0}{0}  $$

$$ \begin{pmatrix} -9 & - 3 \\ -6 & -2 \end{pmatrix}    \binom{x}{y} = \binom{0}{0}  $$


De nuevo tomemos la primera ecuacion

$$-9x - 3 y = 0 $$
$$3x + y = 0 $$
$y = -3x$, o sea que

$$v = \binom{x}{-3x} $$
Tomemos $x=-1$,

$$v = \binom{-1}{3} $$


```python
import numpy as np

A = np.array([[1, 1/2], [1/2, 1]])

# find the eigenvalues/eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```

**Ejemplo 2.2.2:** encuentre los autovalores/vectores de la matriz

$$ A = \begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix}.$$

Explique el significado de estos haciendo el mapeo sobre un cuadrado de esquinas $(1,1), (-1,1), (-1,-1), (1, -1) $
Tambien sobre un circulo unitario $x^2 + y^2 = 1$ (parametricamente se escribe $(\cos \theta, \sin \theta)$, $\theta \in [0, 2 \pi)$.

**Solucion**: El polinomio caracteristico esta dado por

\begin{eqnarray} \det(A - \lambda I) &=&
\det \begin{pmatrix} 1 - \lambda & 1/2 \\ 1/2 & 1 - \lambda \end{pmatrix} \\
&=& (1 - \lambda)^2 - \frac14 \\
&=& (1 - \lambda - 1/2) (1 - \lambda + 1/2) \\
&=& (1/2 - \lambda)(3/2 - \lambda).
\end{eqnarray}
Con esto hay dos valores de $\lambda$, $\lambda_1=3/2$, $\lambda_2=1/2$

Hallemos los autovectores

1. Para $\lambda=3/2$.

$$(1 - 3/2)x + (1/2)y = 0 $$
$$(-1/2)x + (1/2)y = 0 $$
$$x -y = 0 $$
La solucion es  $y=x$,

$$u = \binom{x}{x} $$
Tomemos $x=1$

$$ u = \binom{1}{1}. $$

2. Para $\lambda=1/2$:

$$(1 - 1/2)x + (1/2)y = 0 $$
$$(1/2)x + (1/2)y = 0 $$
$$x + y = 0 $$
o sea que $y=-x$,

$$u = \binom{x}{-x} $$
tomemos $x=1$

$$u = \binom{1}{-1} $$

Apliquemos la matriz $A$ a un cudrado de vertices ....

$$\begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix} \binom{1}{1} = \binom{3/2}{3/2}$$

$$\begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix} \binom{-1}{1} = \binom{-1/2}{1/2}$$

$$\begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix} \binom{-1}{-1} = \binom{-3/2}{-3/2}$$

$$\begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix} \binom{1}{-1} = \binom{1/2}{-1/2}$$
