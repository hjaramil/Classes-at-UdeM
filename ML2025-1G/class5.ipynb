# La pseudoinversa
## Introduccion
Muy pocas (estadisticamente son 0) las matrices que tienen inversa. Todas deben ser cuadradas. Existe el concepto de **pseudoinversa** que inverte todo. Todas las matrices tienen pseudoinversa y es unica. Que es la pseudoinversa?
El problema central del algebra lineal

$$A x  = b. $$
Este problema casi nunca tiene solucion. Una forma de solucionarlo es hallando el minimo de $\| Ax - b \|^2$
(minimos cuadrados). Sin embargo, aunque halla un minimo, este minimo no necesariamente es unico (pueden existir una cantida infinita de soluciones). Si $\mathcal{N} \ne \{ 0 \}$, y $x$ es una solucion entonces $x + x_0$ donde $A x_0 = 0$ ($ x_0 \in \mathcal{A}$)
es solucion pues


$$A (x + x_0) = A x + A x_0 = Ax = b $$.
Si el numero de soluciones $x$ es infinito, entonces
tome la mas pequenha $\min \| x \|$. La pseudoinversa se define como $A^+$ tal que $A^+ x =b$, y $x$ es la solucion obtenida de esta forma.

Para la clase de hoy toca tener claro dos conceptos:

1. Espacio nulo $\mathcal{N}(A)$
1. Espacio columna o rango $\mathcal{R}(A)$.

Observe si $Ax=b$ es entonces $b \in \mathcal{R}(A)$
$f(x) = Ax$, el rango de $f$ es el rango de $A$.

Un comentario, el espacio nulo de una matriz no es el trivial cuando las columnas de $A$ no son linealmente independientes (algebra lineal)


Hagamos el siguiente analisis.

* Si $b \in \mathcal{R}(A)$ entonces la solucion de $Ax=b$ existe. De aca hay dos posibilidades:

    * Si $\mathcal{N}(A) = \{ 0 \}$, entonces $A^{-1}$ existe y $x = A^{-1} b$. En este caso la pseudoinversa ees $A^+= A^{-1}$.

    * Si $\mathcal{N}(A) \ne \{ 0 \}$ la solucion no es unica. Debemos encontrar la mas pequenha de todas $\min || x ||= x^+$.   La pseudoinversa en este caso es $A^+$ tal que $A^+ x^+ = b$.

    Ejemplo: Encuentre la pseudoinversa de $A=(2,3)$.
    Sabemos que el rango es $\mathbb{R}$. Entonces si por ejemplo  $(x,y)^T (2,3) = 5$. Es decir $Au=b$, donde  $u=(x,y)^T$
    $$2x + 3y = 5 $$
    La solucion (existe siempre) no es unica. Hallemos el espacio nulo

    $$2x + 3y = 0 $$
    $$y = -2x/3  $$
    Para $x=-3$, $y=2(3)/3=2$ o sea que

    $$\mathcal{N} = \alpha \binom{-3}{2} $$

    Probemos
    $$2 (-3 \alpha) + 3 (2 \alpha) = -6 \alpha + 6 \alpha=0 $$

    Cualquier solucion de $2x + 3y=5$ le suma una del espacio nulo y sigue siendo solucion.
    Por ejemplo $x=1, y=1$.

    Nueva solucion
    $$\binom{1}{1} + \alpha \binom{-3}{2} =\binom{1 - 3 \alpha}{1  + 2 \alpha} $$

    Inserte esto en $2x + 3y$

    $$2 (1 - 3 \alpha) + 3 (1 + 2 \alpha) = 2 + 3 -  6 \alpha + 6 \alpha = 5  $$

$$\binom{0.76923}{1.15385} $$
Para encontrar esto resuelva $2x + 3y=5$ y necesitamos otra ecuacion $\langle x, y \rangle$, $y \in \mathcal{N}(A)$.

* Si $b \notin \mathcal{R}(A)$
No existe la solucion de $Ax= b$. Lo que debemos hacer es **proyectar** $b$ al rango $\mathcal{R}(A)$. En las notas de clase yo muestro que la proyeccion de $b$ al rango de $A$ se da mediante las ecuaciones normales

$$A^T A x = A^T b $$
Veamos una grafica.

image.png

Si $2x + 3y = 1$ y $2x + 3y = -2$

$$\begin{pmatrix} 2 & 3 \\ 2 & 3 \end{pmatrix} \binom{x}{y} = \binom{-1}{2} $$
La matriz $A$ es $\begin{pmatrix} 2 & 3 \\ 2 & 3\end{pmatrix}$

El rango de $A$ es el cojunto de vectores
$$\alpha \binom{2}{2}  $$

El problema $A^T A x = A^T b$ si tiene solucion.
De aca salen dos casos

* Cuando $A^T A x = A^T b$ es decir cuando proyectamos $b$ al rango de $A$ , $\mathcal{R}(A)$.
    * Si $\mathcal{N}=\{ 0 \}$. La solucion es unica y existe la inversa de $A^T A$, la soluciones

    $$x = (A^T A)^{-1} A^T b $$
    A la matriz $A^+ = (A^T A)^{-1} A^T$ se le denominca la pseudoinversa. Keener dice que esta es la Moore-Penrose pesudoinverse. Yo no comparto esto.

    * Si $\mathcal{N} \ne \{ 0 \}$, entonces $A^T A$ no tiene inversa (esto pasa cuando las columnas de $A$ no son linealmente independientes).
    Este es el peor de todos los casos.
    Se puede hacer lo siguiente. Regularizar (Tikhonov)

    El problema de least squares regularizado es

    $$(A^T A + \lambda I)x = A^T b $$
    $\lambda>0$,
    entonces

    $$x = (A^T A + \lambda I)^{-1} A^T b  $$
    y la pseudoinversa (como es unica)

    $$A^+ = \lim_{\lambda \to 0} (A^T A + \lambda I)^{-1} A^T b $$

image.png

La definicion de Keener de la pseudoinversa es:
La pseudoinversa de la matriz $A$ es la matriz $A^+$ tal que $x = A^+ b$ satisface:

1. $A^T A x = A^T b $ (ecuaciones normales. Proyeccdion de $b$ al rango de $A$, $\mathcal{R}(A)$)
2. $\langle x, w \rangle = 0$ para todo $w \in \mathcal{N}(A)$.

# Algoritmo para la pseudoinversa.
Hay varios algoritmos para encontrar la pseudoinversa (Keener). Vamos a usar el algoritmo de SVD.

$$A = U \Sigma V^* $$
donde $A$ es una matriz $m \times n$, $U$ es una matriz $m \times m$ con columnas $u_i$ autovectores de $A A^*$, y $V$ es una matriz $n \times n$, con columnas autovectores de $A^* A$. $\Sigma$ es diagonal $m \times n$ con elementos en la


$$\Sigma = \begin{pmatrix} \Sigma_r & | & 0 \\ --&--&-- \\ 0 & | & 0  \end{pmatrix}  $$

La pseudoinversa de $\Sigma$ es
las componentes de $\Sigma_r > 0 $,  $\Sigma_r$ es $r \times r$.

$$\Sigma^+ = \begin{pmatrix} \Sigma_r^{-1} &|& 0  \\ --&--&-- \\ 0 & | & 0  \end{pmatrix}  $$

image.png

Analiticamente podemos escribir la pseudoinversa $\Sigma^+$
como

$$\sigma_{ij}^+ =
\begin{cases}
\delta_{ij} \sigma_i^{-1}  , \quad i \le r \\
0 \quad , \quad  i > r
\end{cases} $$

La pseudoinversa de $A$ es

$$A^+ =  V \Sigma^+ U^* $$
La justificacion del algoritmo esta en Keener.
Si $A$ es $m \times n$, $\Sigma^+, A^+$ son $n \times m$.

## Regresion Lineal.
La regresion lineal se resume en una ecuacion

$$X \theta =y . $$
La idea es encontrar (aprender) el mejor $\theta$

**Ejemplo** (minimos cuadrados) el problema de la casa.
Yo quiero vender mi casa y quisiera saber cual es un precio justo. El metodo es:

* Buscar una base de datos de precios de casas vecinas.
* Esta base de datos tambien debe tener parametros. (features)
por ejemplo tamanho de la casa, numero de habitaciones, sector (estrato), arquitectura, etc. Para comenzar solo vamos a usar un "feature". El tamanho de la casa.
* Entrena un modelo de ML. Este le produce $\theta$
* Con $\theta$ usted halla el precio de su casa.

Llevemos esto a un modo mas cuantitativo.
El ejemplo usa la siguiente tabla.



image.png

La idea es un modelo lineal de la forma

$$y_i = \theta_0 + \theta_1 x_i $$
El $\theta_0$ se le llama **bias** (intercepto). El
$\theta_1$ es la **pendiente**.

El modelo seria

$$y_1 = \theta_0 + \theta_1 x_1 $$
$$y_2 = \theta_0 + \theta_1 x_2 $$
$$ \vdots $$
$$y_m = \theta_0 + \theta_1 x_m $$
Este sistema se puede escribir como $y=X \theta$
donde

$$y= \begin{pmatrix} y_1 \\ y_2 \\ \vdots \\ y_m \end{pmatrix} $$
$$\theta = \begin{pmatrix} \theta_0 \\ \theta_1 \end{pmatrix} $$

$$
X =
\begin{pmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots & \vdots \\
1 & x_m
\end{pmatrix}
$$

Vamos a ver como este es un problema de minimos cuadrados.

Usamos la siguiente grafica de los datos



image.png


Vamos a buscar la linea que mejor se ajuste a los datos.

Una forma es minimizando los cuadrados de los errores.
Un error es la diferncia entre el valor $y$ del dato y el valor en la linea azul

$$y_i - \theta_0 - \theta_1 x_i  $$
$$J(\theta) = \sum_{i=1}^m | y_i - \theta_0 - \theta_1 x_i |^2  $$

Se puede escribir como norma

$$J(\theta) = \frac12 \| X \theta - y \|^2 $$
Ya vimos que el gradiente de $J$ (clase anterior) esta dado por

$$\nabla J = X^T X \theta - X^T y $$
Que $\nabla J = 0$ equivale a


$$X^T X \theta = X^T y $$
Y estas son precisamente las ecuaciones normales.
Asi se resuelve el problema de regresion lineal.

Si $X^T X$ es no singular entonces


$$\theta = (X^T X)^{-1} X^T y  $$
de otra forma, regularizamos

$$\theta = (X^T X + \lambda I)^{-1} X^T y  $$
$$\theta = (A + \lambda I)^{-1} b  $$
dondde $A=X^T X$, $b=X^T y$




Esta es una solucion analitica. Debemos contemplar otras soluciones. El metodo iterativo del gradiente.  Se basa en la siguiente ecuacion

$$ \theta^{(i)} = \theta^{(i-1)} - \alpha \nabla J(\theta^{(i-1)})$$
Por que se usa este metodo de iteracion de gradiente? Si la matriz $A$ es muy grande ($1000 \times 1000$) el metodo analitico es muy caro. En la practica se usa el metodo de descenso del gradiente para resolver este problema. Ademas muchas matrices son poco densas (con muchos ceros, sparse) y el metodo del gradiente es mas efectico aca.

El gradiente muestra el camino de maximo ascenso. El negativo del gradiente muestra el camino del maximo descenso. El $\alpha$ es un parametro para calibrar que tan rapido descendemos o no. Sin embargo un $\alpha$ muy grande nos puede mandar a la direccion opuesta y un $\alpha$ muy pequenho nos puede hacer descender muy lentamente.
