# Ecuaciones Hiperbolicas
La tipica ecuacion hiperbolica es la **ecuacion de onda**.
Primero resolvemos la ecuacion de onda en una dimension (espacial). Una cuerda o una barra.
Esta esta dada por

$$
\begin{cases}
\frac{\partial^2 u}{\partial x^2} - \frac{1}{c^2(x)} \frac{\partial^2 u }{\partial t^2} = - g(x,t) \\
u(a,t) = e(t) \quad , \quad u(b,t) = f(t) \quad , \quad \text{condiciones de frontera} \\
u(x, t_0) = p(x) \quad , \quad u_t(x,t_0)= q(x) \quad , \quad x \in [a,b] \quad , \quad \text{condiciones iniciales}
\end{cases}
$$

Queremos encontrar el campo de onda $u(x,t)$, $t>t_0, x \in [a,b] $
$t$ es finito. Las funciones $e,t$ son de frontera y las $p,q$ iniciales. $g(x,t)$ es la fuente.

El esquema que vamos a plantear es central en espacio y en tiempo

$$\frac{w_{i+1 j}- 2 w_{ij} + w_{i-1 j}}{\Delta x^2} = \frac{1}{c_i^2} \frac{w_{i j+1} - 2 w_{ij} + w_{ij-1}}{\Delta t^2}  - g_{ij} \tag{1} $$
con error de $\mathcal{O}(\Delta x^2)$ y $\mathcal{O}(\Delta t^2)$. Aca

$$g_{ij} = g(x_i, t_j), c_i = c(x_i) ,  x_i = x_0 + i \Delta x, t_j = t_0 + j \Delta t $$

**Condiciones iniciales y de frontera**
$$ w_{0j} = e(t_j)=e_j  \quad , \quad w_{nx j} = f(t_j) = f_j$$
$$w_{i0} = p(x_i) = p_i $$
Para el caso $u_t(x,t_0) = q(x)$ tenemos  central en 0.

$$ \frac{w_{i1}- w_{i-1}}{2 \Delta t} = q(x) + \mathcal{O}(\Delta x^2)$$
Esto lo usamos por que necesitamos $w_{i-1}$?
Despejemos esto


$$w_{i \; -1} = w_{i1} - 2 q_i \Delta t \tag{2} $$

Como es costumbre vamos a utilizar una variable adicional
$$\gamma_i = \frac{c_i \Delta t}{\Delta x} \quad , \quad c_i = c(x_i) $$



De la Ecuacion (1), despejando $w_{i j+1}$

$$w_{i \; j+1} = \gamma_i^2 ( w_{i+1 j} - 2 w_{ij} + w_{i-1 j}) + 2 w_{ij} - w_{i j-1} + g_{ij}  c^2 \Delta t^2  $$

$$w_{i \; j+1} = \gamma_i^2 w_{i-1 j} + 2(1 - \gamma_i^2) w_{ij} + \gamma_i^2 w_{i+1 j} - w_{i j-1} + g_{ij} c_i^2 \Delta t^2
\tag{3} $$

## Analisis de Estabilidad.
Recuerden que la idea es que el factor de ganancia

$$G = \frac{ w_{i j+1}  }{w_{ij} } \le 1$$

El metodo de John von Neumann asume que la solucion es una superposicion de funciones elementales de la forma

$$u_{\omega k} = C(\omega, k) \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x} \tag{4} $$ Sin perdida de generalidad asumimos que $C(\omega, k)=1$, por que en $G$ se cancela.   $\omega=2 \pi f$, donde $f$ es la frecuencia. $\omega$ es la frecuencia cirular. $k$ se llama el numero de onda $k = 2 \pi/ \lambda$, donde $\lambda$ es la longitud de onda. Hay un paralelo

$$\omega = \frac{2 \pi}{T}  $$
$$k  = \frac{2 \pi}{\lambda}  $$
Asi como $T$ es el periodo en tiempo, $\lambda$ el periodo espacial.

### Relacion de dispersion.
Si insertamos la ecuacion (4) en la ecuacion de onda, que encotramos?

Veamos

$$\frac{\partial^2 u}{\partial x^2} = -C(\omega, k) k^2 \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x}  $$

$$\frac{\partial^2 u}{\partial t^2} = -C(\omega, k) \omega^2 \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x}  $$

De la ecuacion de onda queda

$$ \frac{\partial^2 u}{\partial x^2}  - \frac{1}{c^2(x)} \frac{\partial ^2 u}{\partial t^2}  $$

$$-C(\omega  , k) k^2   \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x}  + \frac{1 }{c^2(x)} C(\omega, k)  \omega^2 \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x}= 0 $$

Dividiendo por $C(\omega, k)$, y por $ \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x}$
encontramos


$$\frac{\omega^2}{c^2(x)} - k^2 = 0 $$

De forma que

$$k = \pm \frac{\omega^2}{c(x)} $$
Encontramos que las variables $k,\omega$ son independientes. A esto se le llama la **relacion de dispersion** de la ecuacion de onda.

El analisis de estabilidad asume que

$$w_{ij} =  \mathrm{e}^{\mathrm{i} \omega t} \mathrm{e}^{\mathrm{i} k x} $$

De forma que

$$G = \frac{w_{i j+1}}{w_{ij}} =  \frac{\mathrm{e}^{\mathrm{i} \omega t_{j+1}} \mathrm{e}^{\mathrm{i} k x_i}}{ \mathrm{e}^{\mathrm{i} \omega t_j} \mathrm{e}^{\mathrm{i} k x_i}} = \mathrm{e}^{\mathrm{i} \omega \Delta t}
$$

De otro lado si insertamos esta solucion $w_{ij}$ en le ecuacion (3)

$$w_{i \; j+1} = \gamma_i^2 w_{i-1 j} + 2(1 - \gamma_i^2) w_{ij} + \gamma_i^2 w_{i+1 j} - w_{i j-1} + g_{ij} c_i^2 \Delta t^2
$$





Asumimos que la fuente es 0

$$\mathrm{e}^{\mathrm{i} ( w t_{j+1} + k x_i)} = \gamma_i^2 \mathrm{e}^{\mathrm{i} (w t_j + k x_{i-1})} + 2 (1 - \gamma^2) \mathrm{e}^{\mathrm{i} (w t_j + k x_i)}
+ \gamma_i^2 \mathrm{e}^{\mathrm{i} (w t_j + k x_{i+1})} - \mathrm{e}^{\mathrm{i}(w t_{j-1} + k x_i)} $$




Dividmos todo por $w_{ij}=\mathrm{e}^{\mathrm{i} ( \omega t_j + k x_i)}$

Queda

$$G = \mathrm{e}^{\mathrm{i} \omega \Delta t} =
\gamma^2 \mathrm{e}^{-\mathrm{i} k \Delta x} + 2(1 - \gamma^2) + \gamma^2 \mathrm{e}^{\mathrm{i} k \Delta x} - \mathrm{e}^{-\mathrm{i} \omega \Delta t} = 2 \gamma^2 \cos k \Delta x + 2(1 - \gamma^2) - G^{-1} $$

Obtenemos multiplicando por $G$

$$G^2 = 2(1 - \gamma^2) G + 2 \gamma^2 G \cos k \Delta x - 1 \tag{5}$$

Hagamos la reduccion trigonometrico/algebraica. Recuerde el coseno en terminos del angulo medio.

Tomemos los terminos con coeficiente 2

\begin{eqnarray}
2 (1 - \gamma^2) + 2 \gamma^2 \cos k \Delta x &=&
2[ (1 - \gamma^2) + \gamma^2 \cos k \Delta x]  \\
&=&
2 \left [  (1 - \gamma^2) + \gamma^2 \left ( 1 -2 \sin^2 \frac{k \Delta x}{2}   \right ) \right ]  \\
&=& 2 \left [  1 - 2 \sin^2 \left ( \frac{k \Delta x}{2} \right ) \right ]
\end{eqnarray}
Definamos la variable intermedia

$$\beta = 1 - 2 \gamma^2 \sin^2 \left ( \frac{k \Delta x}{2} \right ) \tag{6} $$
de forma que simplificando (5) ,

$$
G^2 - 2 \beta G + 1 = 0
$$

Resolvemos la cuadratica

$$G_{\pm} = \frac{2 \beta \pm \sqrt{4 \beta^2 - 4}}{2}=\beta \pm \sqrt{\beta^2 - 1} $$

Recuerden que queremos $|G | \le 1$. Si $\beta \ge 1$ entoncces $G_+ > 1$.
Entonces para que $|G | \le 1$ necesitamos que $\beta < 1$. La raiz se vuelve compleja y escribimos

$$G_{\pm} = \beta \pm \mathrm{i} \sqrt{1 - \beta^2}  $$

Cual es la magnitud de $G$ ahora?

$$| G |^2 = G \overline{G} = \beta^2 + (1 - \beta^2) = 1 $$
Esta bien, pues $| G|=1$.

Es decir, de la Ecuacion (6)

$$\left |  1 - 2 \gamma^2 \sin \left ( \frac{ k \Delta x}{2}  \right )   \right | \le 1 $$
Ya hemos resuelto esta desigualdad varias veces y econtramos


$$ 2 \gamma^2 \sin^2 \left ( \frac{k \Delta x}{2}  \right ) \le 2 $$
como $| \sin x | \le 1$, entones es suficiente que

$$\gamma \le 1 $$
Es decir

$$ \frac{c \Delta t}{\Delta x} $$
y esta es la condicion de estabilidad. Llamada
**CFL condition** (Courant-Friedrichs-Lewy).

Esta condicion tiene una interpretacion fisica.

$$\frac{c \Delta t}{\Delta x} \le 1 $$

La idea es que

$$c \le \frac{\Delta x}{\Delta t} \tag{7} $$



image.png

La ecuacion (7) nos dice, busque $\Delta x$, $\Delta t$ tales que la velocidad numerica $\Delta x/\Delta t$ sea mayor que la velocidad de propagacion $c$.

Digamos que en este analisis la velocidad se considero contante $c$? que pasa si la velocidad es variable?

Con velocidad constante tenemos la siguiente relacion para que halla estabilidad

$$\Delta t \le \frac{\Delta x}{c} $$
Para velocidad variable escoja

$$C = \max \{ c_i \} \tag{8} $$
y
$$\Delta t \le \frac{\Delta x}{C} $$

Asumamos
$p(x) = 6 \sin \left ( \frac{\pi x}{L} \right ) $
$x \in [ 0 , L]$, $L=1$, $c=1$, $t_0=0$, $a=0$, $b=L=1$.

```python
from re import T
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi, sin, cos, array
from matplotlib.pyplot import figure
figure(figsize = (16,10))



# definicion de parametros
a = 0
b=1
t0=0
L=b
dt=0.1
c=1
nx=101

# solucion analitica
def u(x,t):
    return 6*sin(pi*x/L)*cos(c*pi*t/L)

#graficamos las columnas de w
params={'legend.fontsize':5, 'legend.handlength':1}
Ndt=11

# labels and title
plt.xlabel(r"$x$")
plt.ylabel(r"desplazamiento")
plt.title(r"Curvas de Onda, Solucion analitica")

X = np.linspace(a,b,nx)
for j in range(Ndt):
    t = j*dt
    plt.plot(X, u(X,t), label=str( round(t,1)) + '  seg')

plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., title='Tiempo (s)')
plt.show()
```

```python
# implementacion del algoritmo
def solve_waveeq(c, a, b, t0, nt, nx, dt, e, f, p, q, g):
    X = np.linspace(a,b,nx)
    cmax = max(c(xx) for xx in X)

    L=b-a
    dx=L/(nx-1)
    gamma=cmax*dt/dx
    stab_cond=dx/cmax

    # test stability
    if dt >= stab_cond:
        print("error: la condicion de estabilidad fue violada. Reduzca dt")
        # return None

    tmax=nt*dt
    t=np.arange(t0,tmax,dt)

    w=np.zeros((nx,nt),dtype=float)
    print("dimensions of w", w.shape)

    # condiciones iniciales
    for i in range(1,nx-1):
        w[i,0]=p(X[i])

    # condiciones de frontera
    for j in range(nt):
        w[0,j] = e(t[j])
        w[nx-1, j]=f(t[j])

    # lleno de capa 1
    for i in range(1, nx-1):
        gamma = c(X[i]) * dt / dx
        w[i,1] = (0.5*gamma**2*p(X[i-1]) + (1.0-gamma**2)*p(X[i]) +
                  0.5*gamma**2*p(X[i+1]) + q(X[i])*dt
                  + 0.5*c(X[i])**2**dt**2*g(X[i], t0))

        # como ya se tienen dos capas se puede iterar en tiempo
    for j in range(1, nt-1):
        for i in range(1, nx-1):
            gamma = c(X[i]) * dt / dx
            w[i, j+1] = (gamma**2 * w[i-1, j] + 2.0*(1.-gamma**2) * w[i, j] +
                          gamma**2 * w[i+1, j] - w[i,j-1]
                          + g(X[i], t[j])*c(X[i])**2*dt**2)

    return w
```

```python
# prueba numerica
# velocidad
def v(x):
    return 1

# condiciones de borde 0
def e(x):
    return 0

def f(x):
    return 0

# condicion de borde en L
def p(x):
    return 6*sin(pi*x/L)

def q(x):
    return 0

# definicion de la fuente
def g(x,t):
    return 0

dt = 0.001
nt = 1001

w = solve_waveeq(v, a, b, t0, nt, nx, dt, e, f, p, q, g)
```

```python
# graph the solution
params={'legend.fontsize':5, 'legend.handlength':1}
Ndt=11

# labels and title
plt.xlabel(r"$x$")
plt.ylabel(r"desplazamiento")
plt.title(r"Curvas de Onda, Solucion numerica")

X =np.linspace(a,b, w.shape[0])
Ndt =110

for j in range(0, nt, Ndt):
    t = j*dt
    plt.plot(X, w[:, j], label=str(round(t,1)) + '  seg')
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., title='Tiempo (s)')

plt.grid(True)
plt.show()
```
