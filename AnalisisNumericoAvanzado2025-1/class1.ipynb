## Tecnicas iterativas en el algebra lineal.
Recordemos los metodos para resolver ecuaciones lineales. El de Gauss (eliminacion Gaussiana) es efectivo pero si las matrices son inmensas y con muchos ceros (esparsas). Los metodos iterativos vienen al rescate. Son muy rapidos y consumen poca memoria para matrices esparsas.

Los metodos iterativos se basan en la ecuacion


$$\bmx_i = T \bmx_{i-1} +  \bmc$$
Donde $\bmx_i$ es el valor actual (presente,hoy), $\bmx_{i-1}$ es el valor anterior (ayer) y $\bmc$ es una constante.

Por ejemplo, comenzamos con el "initial guess", solucion inicial $\bmx_0$, entonces creamos una cadena
`
$$\bmx_1 = T \bmx_0 + \bmc $$
$$\bmx_2 = T \bmx_1 + \bmc $$
$$\vdots $$
$$\bmx_n = T \bmx_n-1 + \bmc $$
$$\vdots $$
Lo importante es que el sistema converja. Que quiere decir **converja**?
Que la secuencia

$$S = \{ \bmx_0, \bmx_1, \cdots , \bmx_n, \cdots  \} $$
Tenga un limite. Que $\bmx = \lim_{n \to \infty} \bmx_{n}$.
Si existe tal limite entonces

$$\bmx = T \bmx + \bmc  $$

En este sentido $\bmx$ es un **punto fijo**. Si no se ve asi piense
en otro operador $S \bmx = T \bmx + \bmc$ de forma
en el limite
$$S \bmx = \bmx $$
$\bmx$ es punto fijo del operador $S$.

Vamos a estudiar varios metodos iterativos que son utiles para este curso, en la solucion de sistemas lineales:
* Jacobi
* Gauss-Seidel
* Over-relaxation (sobre-relajacion)

### Motivacion.
Pensemos en la distancia entre dos vectores en $\mathbb{R}^2$. Una forma de definir esta distancia es por la metrica **pitagorica**:
Sean $(x_1,x_2) \in \mathbb{R}^2$
 $(y_1,y_2) \in \mathbb{R}^2$. La distancia pitagorica entre ambos esta dada por la formula

 $$d[ (x_1, x_2), (y_1, y_2) = \sqrt{ (x_1-y_1)^2 + (x_2-y_2)^2} $$
 De igual forma esta ecuacion se puede generalizar a $\mathbb{R}^n$ como\

 $$d[ (x_1, x_2, \cdots, x_n), (y_1, y_2, \cdots, y_n)] = \sqrt{ \sum_{i=1}^n (x_i - y_i)^2}$$

 Definimos ahora la norma de un vector de forma **axiomatica**
 Sea $\bmx \in \mathbb{R}^n$, si existe una funcion $\| . \|:\mathbb{R}^n \to \mathbb{R}^n \cup \{ 0 \}$ tal que


 1. $\| \bmx \| \ge 0$, y $\| x \| = 0$ si y solo si $\bmx = 0$. **positividad**

 2. $\| \alpha \bmx \| = | \alpha | \| \bmx \|$, **homogeneidad**
 3. $\| \bmx + \bmy \| \le \| \bmx \|  + \| \bmy \|$
 Se dice que la funcion $\| \bmx \|$ es la norma del vector $\bmx$.

 **Ejemplos:**:
 La norma mas popular es la norma $p$. Se llama $\ell_p$. Se define con la formula

 $$\ell_p(\bmx) = \| \bmx \|_p = \left ( \sum_{i=1}^n |x_i|^p   \right )^{1/p}  \quad , \quad p \ge 1  $$

 Ejemplos particulares:

 1. $p=1$:
 $$\ell_1(\bmx)= \| \bmx \|_1 = \sum_{i=1}^n | x_i |. $$
 Se puede probar que esta norma satisface los axiomas de la definicion de arriba.

 2. $p=2$:
 $$\ell_2 (\bmx)= \| \bmx \|_2 = \sqrt{ \sum_{i=1}^n | x_i |^2} $$
 Esta es la norma **pitagorica**

3. $p=\infty$ se llama la norma del maximo

$$\ell_{\infty}(\bmx) = \| \bmx \| = \max | x_i | $$

Dejamos como ejercicio probar que

$$\lim_{p \to \infty} \ell_p = \ell_{\infty} $$




Existe una ditancia **inducida** por la norma. Por ejemplo

$$d(x,y) = \ell_p( \bmx - \bmy) = \| \bmx - \bmy \|_p = \left (  \sum_{i=1}^n | x_i - y_i|^p \right )^{1/p} $$



Tambien existe la norma de una matriz. Vamos a definir la norma de una matriz que ademas debe satisfacer la relacion

$$ \| A B \| \le \| A \| \| B \| $$
y tambien podemos pensar en la **distancia entre dos matrices** como
$d(A, B) = \| A - B \|$. Definimos entonces enseguida la norma de una matrix.

Sea $A \in \mathbb{R}^{m \times n}$  (tambien puede ser $\mathbb{C}^{m \times n}$, definimos a norma de $A$ como


$$ \| A \| = \sup \{ \| A x \| : \| x \| = 1   \}  = \sum \left \{ \frac{\| A \bmx \|}{ \| \bmx \|  } : \bmx \ne 0  \right \}$$
Explico a mano alzada el concepto de "sup".

Donde $\bmx $ y $ A \bmx $ se asumen en espacios $\mathbb{R}^n$ y $\mathbb{R}^m$

De esta definicion se pueden encontrar las siguientes normas de matrices.

1. Normas $p$. Para $p=1$

$$\| A \|_1 = \max_{1 \le j \le n}  \sum_{i=1}^m | a_{ij} | $$
En palabras: El maximo de la suma por columnas.  Ejemplo a mano alzada.

2. La norma $p=2$ es la norma **espectral**:
$p=2$

$$ \| A \|_2 = \sigma_{max}(A) $$
donde $\sigma_{max}(A)$ es el valor singular mas grande.
Que es **valor singular**? enseguida vamos a ver la definicion de autovalor. Los valores singulares son las raices cuadradas de los autovalores de $A^T A$.
Queda pendiente la clarificacion de esta definicion por que depende del concepto de **autovalor** el cual no hemos definido todavia.

3. $p=\infty$

$$\| A \|_{\infty} = \max_{1 \le i \ln n} | a_{ij} | $$
En palabras: seria el maximo por filas. Ejemplo a mano alzada.

4. **norma Frobenius**:

Para matrices cuadradas $n \times n$
$$\| A \|_F = \sqrt{ \sum_{i=1}^n \sum_{j=1}^n | a_{ij}^2 | }$$
Ejemplo a mano alzada.

Se deja como ejercicio probar como para $p=1,2, \infty$ se obtienen estas normas $p$ de $\| A \|_p = \sup \{ \| A \bmx \|_p : \| x \|_p =1 \}$

Pasamos ahora a ver dos conceptos muy importantes en el algebra lineal y el ML(Machine Learning)

**Definicion AutoValor-AutoVector**. Sea $A$ una matriz cuadrada (compleja o real) si existe un escalar $\lambda$ y un vector $\bmx$ tales que

$$A \bmx = \lambda \bmx $$
decimos que $\lambda$ es un **autovalor de la matriz $A$ y que $\bmx$ es el **autovector** correspondiente a $\lambda$ de la matriz $A$.

Veamos la interpretacion de este concepto, a mano alzada y Wikipedia.


El concepto de autovalor/vector es muy importante en  varias areas del conocimiento:

* Mecanica cuantica. Los estados de energia son autovalres del operador de Schrodringer (la ecuacion de Scrodinger)

* En la teoria de la elasticidad. Los autovalores son frecuencias de estados de vibracion.

* En la optica. El arco-iris es una descomposicion de la luz blanca en el **espectro** de autovalores.  La palabra **autovalor** se asocia con espectro.
