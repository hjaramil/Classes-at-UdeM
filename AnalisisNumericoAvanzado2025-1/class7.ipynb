# Empalme con la clase anterior
En general, asumamos $n$ variables independientes $x_1, x_2, \cdots, x_n$, la variable escalar dependiente
$u=u(x_1, x_2, \cdots, x_n)$. Definimos en general la derivada parcial de orden $i_1 + i_2 + \cdots + i_n$
como

$$\frac{\partial^{i_1+i_2+ \cdots + i_n} u }{\partial x_1^{i_1} \partial x_2^{i_2} \cdots \partial_{x_n}^{i_n} } = u_{x_1^{i_1} x_2^{i_2} \cdots x_n^{i_n}} = u_{,{1^{i_1} 2^{i_2}, \cdots n^{i_n}}} $$
Done $i_j=0,1,2, \cdots, $, $j=1,2, \cdots, n$.


El grado $i_1 + i_2 + \cdots i_n \ge 1$, sino, no hay ecuacion diferencial (algebraica).


Por ejemplo que pasa si $i_1=1$, $i_2=2$, $i_3=0$ tres variables $x_1, x_2, x_3$

$$\frac{\partial^{i_1+i_2+ \cdots + i_n} u }{\partial x_1^{i_1} \partial x_2^{i_2} \cdots \partial_{x_n}^{i_n} } =
\frac{\partial^3 u}{\partial x_1 \partial x_2^2}
= \frac{\partial^3 u}{\partial x \partial y^2}= u_{x_1 x_2^2} = u_{x y^2} = u_{, 12^2} = u_{,122}$$

Las ecuaciones diferenciales en derivadas paciales en las variables $x_1, x_2, \cdots, x_n$ para la solucion integral $u=u(x_1, x_2, \cdots, x_n)$
se escriben, en general

$$f \left ( x_1, x_2, \cdots, x_n, u,  \frac{\partial^{i_1+i_2+ \cdots + i_n} u }{\partial x_1^{i_1} \partial x_2^{i_2} \cdots \partial_{x_n}^{i_n} }\right ) = 0. $$

Dentro de la gama de ecuaciones diferenciales parciales tenemos las **lineales** de segundo orden que son las que tratamos en esta clase.

$$\sum_{i,j=1}^n A_{ij} u_{,ij} + \sum_{i=1}^n B_i u_{,i} + g(x_1, x_2, \cdots, x_n) = 0 $$

Especial aca es que $A_{ij}$ no puede depender ni de $u$ ni de ninguna de sus derivadas parciales. Lo mismo con $B_i$.

Asuma que si $u$ es una funcion de las variables $x,y,t$. En aplicaciones comunies $t$ es tiempo. Es decir, $u=u(x,y,t)$. Consideramos la siguiente ecuacion

$$A u_{,11} + B u_{,12} + C u_{,22} + D u_{,13} + E u_{,23} + F u_{,33} + G_{u,1} + H u_{,2} + I u_{,3} + g(x,y,t) = 0 $$

Asumamos el orden $x,y,t$

$$A u_{xx} + B u_{xy} + C u_{yy} + D u_{xt} + E u_{yt} + F u_{tt} + G_{u_x} + H u_{_y} + I u_{_t} + g(x,y,t) = 0 $$

Vamos a cambiar un poco la notacion.

$$
X = \frac{\partial}{\partial x}   \quad , \quad
Y = \frac{\partial}{\partial y}   \quad , \quad
T = \frac{\partial}{\partial t}
$$
Es conveniente pensar en estas variables como si fueran algebraicas (son diferenciales).

Con esto vamos a escribir las ecuaciones **canonicas** de las ecuaciones diferenciales parcicales lineales de segundo orden

Noten que

$$X^2 = \frac{\partial }{\partial x} \frac{\partial}{\partial x} = \frac{\partial^2}{\partial x^2}  $$

1. **Elipticas**:
El discriminante es $<0$

$$f(X,Y) = \frac{X^2}{a^2} + \frac{Y^2}{b^2} - 1 = 0 $$

Esto lo podemos escribir como

$$A(x,y) X^2 + B(x,y) Y^2 = g(x,y) $$
donde $A>0, B>0$. A esta ecuacion se le llama eliptica.
A veces escribimos la fuente $g(x,y)=\rho(x,y)$. Si $A=B=1$
tenemos un caso especial


$$ \nabla^2 u \equiv  \frac{\partial^2 u}{\partial x^2}  + \frac{\partial^2 u}{\partial^2 y} = \rho(x,y) $$
A esta ecuacion se le conoce como la ecuacion de **Poisson**
en dos variables.

Otro caso especial
$$ \nabla^2 u \equiv  \frac{\partial^2 u}{\partial x^2}  + \frac{\partial^2 u}{\partial^2 y} = 0 $$
se le conce como la ecuacion de **Laplace**.

2. **Parabolicas**:


$$f(X,T) = T - \alpha X^2 - g(x,y) = 0. $$
La ecuacion de **calor**

$$ \frac{\partial u}{\partial t}  - \alpha \frac{\partial^2 u}{\partial x^2} = g(x,t) $$



3. **Hiperbolicas**
$$f(X,T) = \frac{X^2}{a^2} - \frac{T^2}{b^2} = 1 $$

Podemos escribir

$$A(x,t) \frac{\partial^2 u}{\partial x^2} - B(x,t) \frac{\partial^2 u}{\partial x^2}  = g(x,t) $$
$$A(x,t)>0 \quad , \quad B(x,t)>0$$.

Caso especial $A=1$, $B(x,t)= \frac{1}{c^2(x)}$
esta es la ecuacion de onda en una dimension espacial $x$


$$ \frac{\partial^2 u}{\partial x^2}  - \frac{1}{c^2(x)} \frac{\partial^2 u}{\partial t^2} = g(x,t). $$

En general para funciones de dos variables podemos escribir la ecuacion

$$A u_{xx} + B u_{xy} + C u_{yy} + D u_x + E u_y + F u + G = 0. $$
El discriminante es

$$Discriminante= \sqrt{B^2 - 4 AC} $$
Si  $Discriminante < 0$ eliptica

Si  $Discriminante = 0$ parabolica

Si  $Discriminante >0 $ hiperbolica

# Diferencias Finitas:
Metodo para resolver ecuaciones diferenciales numericamente.
El tratamiento que damos aca es limitado. La referencia Smith

[9] G. D. Smith. Numerical Solution of Partial Differential Equations: Finite Difference
Methods. 1985.

Es mas completa por mas adelante quieren profundizar sobre el tema.

Debemos definir el dominio. Los valores posibles de $u(x,y)$ (en $\mathbb{R}^2$), al final consideramos el caso de un dominio en $\mathbb{R}^3$. Por ahora dos dimensional.

1. El dominio de $u$ esta definido por un rectangulo
$$R,  a \le x \le b  \quad , \quad c \le y \le d $$

La frontera la especificamos como

$$ B = \{ (x,y) | (x = a  \; \lor \; x=b)  \land (y=c \lor y=d )  \} $$

Desde la frontera se extrapola el campo a los puntos interiores.

2. Se crea la **malla** (grid). Esta esta conformada por los puntos $(x_i, y_i)$ tales que

* $x_0=a, y_0=c, x_i= x_0 + i h, y_j=y_0 + jk$
* Con $h=(b-a)/n$, $k=(d-c)/m$ $b>a$, $d>c$,
* $i=0.1, \cdots, n$, $j=0,1, \cdots, m$.

3. La solucion se denota como $u_{ij} = u(x_i, y_j)$. Es decir, la solucion es una matriz con el valor solucion en cada punto de malla.

4. Hay condiciones iniciales (cuando se usa timempo, $t_0$) y de frontera. Las de frontera son en $(x_0, y_j)$, $(x_n, y_j)$, en $(x_i, y_0)$, $(x_i, y_m)$.  En estos puntos se conoce la solucion.
Es comun que las de arriba $(x_i, y_m)$ no se necesiten excepto en las ecuaciones elipticas.

5. La solucion $u=u(x,y)$ se llama analoga (esta se estudia en el curso de PDE). Aca se establecen soluciones aproximadas usando diferencias finitas.

## Derivadas de primer orden.
Usando la aproximacion de Taylor

$$u_x(x,y) = \frac{u(x+h,y) - u(x,y) }{h} - u_{xx}(c,y) \frac{h}{2} $$
donde $c \in [x, x+h]$, $h>0$. Para los puntos interiores de la malla tenemos

$$u_x(x_i,y_j) = \frac{u(x+h,y) - u(x_i,y_j) }{h} - u_{xx}(c_i,y_j) \frac{h}{2} $$
con $c_i \in [ x_i, x_i + h]$, $h>0$.
Simplificando escribimos

$$ u_x(x_i, y_i ) \approx \frac{w_{i+1 j} - w_{ij}}{h} \tag{1} $$
donde el error es $\mathcal{O}(h)$. Ojo llamaremos $w_{ij}$ la aproximacion de $u_{ij}$ por que es una aproximacion.


A la ecuacion (1) se llama primera diferencia hacia adelante. (forward finite difference)
Existen otras versiones

$$ u_x(x_i, y_i ) \approx \frac{w_{i j} - w_{i-1 \, j}}{h} \tag{2} $$
La (2) se llama diferencias finitas hacia atras.
Las diferencias (1) y (2) son de orden $\mathcal{O}(h)$

$$ u_x(x_i, y_i ) \approx \frac{w_{i+1 j} - w_{i-1 \, j}}{2h} \tag{3} $$
 A la (3) se le conoce como diferencias centradas.
La diferencias (3) es de orden $\mathcal{O}(h^2)$

## Derivadas de segundo orden
Usando las tecnicas de arriba (Taylor)

$$u_{xx}(x_i, y_j) \approx \frac{w_{i+1 \, j } - 2 w_{ij} + w_{i-1 \, j}}{k^2} $$

$$u_{yy}(x_i, y_j) \approx \frac{w_{i \, j+1 } - 2 w_{ij} + w_{j-1 \, j}}{k^2} $$
con $k = \Delta y$.

Comenzamos con las parabolicas por que son mas simples, pues la derivada en tiempo es de primero orden. Despues las hiperbolicas y al final las elipticas.

## Ecuaciones Parabolicas.
Escribamos la ecuacion, con doniciones iniciales y de frontera
IBVP: Initial Boundary Value Problem


\begin{cases}
\frac{\partial u}{\partial t}  - \alpha \frac{\partial^2 u}{\partial x^2} = 0  \tag{4} \\
u(x, t_0=c) = I_o(x) \quad , \quad \text{condicion inicial} \\
u(a,t) = h_1(t) \quad, \quad h(b,t) = h_2(t) \quad , \quad \text{condiciones de frontera}
\end{cases}
Aca $\alpha$ se llama **difusividad termica**. Mide que tan rapido se enfria el materia.

### FTCS: Progresivas (hacia adelante) en tiempo y centradas en espacio.

La ecuacion de la primera derivada progresiva en tiempo es

$$u_t(x_i, t_j) = \frac{w_{i j+1} - w_{ij} }{k} + \mathcal{O}(k) \tag{5} $$
donde $k=\Delta y$.

La ecuacion de la segunda derivada central en espacio es

$$u_{xx}(x_i, t_j) = \frac{w_{i+1 j} - 2 w_{ij} + w_{i-1 j}}{h^2} + \mathcal{O}(h^2) \tag{6} $$

Juntando las (4) y (5)  en la ecuacion de calor (4)

$$\frac{w_{i j+1} - w_{ij}}{k}  - \alpha \frac{w_{i+1 j} - 2 w_{ij} + w_{i-1 j} }{h^2} $$
con un error de $\mathcal{O}(k)$ ein tiempo  y $\mathcal{O}(h^2)$ en espacio.

Dspejamos el termino en $j+1$ de aca

$$w_{i j+1} = w_{ij} + \frac{\alpha k}{h^2} ( w_{i+1 j} - 2 w_{ij} + w_{i-1} j), $$
Agrupamos para juntar los $w_{ij}$

$$w_{i j+1} = \left (  1 - 2 \frac{\alpha k}{h^2}  \right ) w_{ij} + \frac{\alpha k}{h^2} ( w_{i+1 j} + w_{i-1 j} )  $$

o de forma simplificada

$$ w_{i j+1} = (1 - 2 \gamma) w_{ij} + \gamma ( w_{i+1 j} + w_{i-1 j} ) \tag{7} $$
donde
$$\gamma = \frac{\alpha k}{h^2}  $$

Debemos analizar esto en la malla de puntos $(i,j)$
$i=0,1, \cdots, n$, $j=0, \cdots, m$

La siguiente figura ilustra lo que se llama **estrella, plantilla, o molecula** del sistema.

El sistema se basa ahora en la ecuaicion (7)
$$ w_{i j+1} = (1 - 2 \gamma) w_{ij} + \gamma ( w_{i+1 j} + w_{i-1 j} )$$

Veamos como se propaga el campo

Las "paredes" las conocemos

$$ w_{i 1} = (1 - 2 \gamma) w_{i0} + \gamma ( w_{i+1 0} + w_{i-1 0} )  \quad , \quad 0 < i < n$$

* $j=1$
$$ w_{i 2} = (1 - 2 \gamma) w_{i1} + \gamma ( w_{i+1 1} + w_{i-1 1} )  \quad , \quad 0 < i < n$$
Las "paredes" las conocemos
$w_{02} = h_1(t_2) \quad, \quad w_{n2} = h_2(t_2) $

$\vdots$
en general

* $j=j$
$$ w_{i j+1} = (1 - 2 \gamma) w_{ij} + \gamma ( w_{i+1 j} + w_{i-1 j} ) \quad , \quad 0 < i < n  \quad , 0 \le j \le m-1$$
Las "paredes" las conocemos
$w_{0j+1} = h_1(t_{j+1}) \quad, \quad w_{n2} = h_2(t_{j+1}) $

Esta ultima ecuacion la podemos escribir en forma matricial
(producto interno, o de dos vectores)

$$w_{1 j+1} = ( \gamma \quad 1 - 2 \gamma \quad \gamma)
\begin{pmatrix}  w_{i-1 j} \\ w_{ij} \\ w_{i+1 j} \end{pmatrix}  \quad , 0 < i < n \tag{8} $$

La ecuacion (8) se puede esacribir hacien i=1,2, \cdots, n-1$ explicitamente como

$$w_{1 j+1} = ( \gamma \quad 1 -2 \gamma \quad \gamma )
\begin{pmatrix}
w_{0j} \\ w_{1j} \\ w_{2j}
\end{pmatrix}
= (1 - 2 \gamma \quad \gamma) \binom{w_{1j}}{w_{2j}} + \gamma w_{0j} $$


$$w_{2 j+1} = ( \gamma \quad 1 -2 \gamma \quad \gamma) \begin{pmatrix} w_{1j} \\ w_{2j} \\ w_{3j} \end{pmatrix}  $$

$$ \vdots $$

$$w_{n-1 j+1} = ( \gamma \quad 1 - 2 \gamma \quad \gamma )
\begin{pmatrix}
w_{n-2 j} \\
w_{n-1 j} \\
w_{n j} \\
\end{pmatrix}
= (\gamma \quad 1- 2 \gamma)
\binom{w_{n-2 j}}{w_{n-1 j}} + \gamma w_{nj} $$


Llevamos todas estas ecuaciones a una forma **matricial**


$$
\begin{pmatrix}
w_{1 j+1} \\
w_{2 j+1} \\
\vdots \\
w_{n-2 j+1} \\
w_{n-1 j+1}
\end{pmatrix}
=
\begin{pmatrix}
1 -2 \gamma & \gamma & 0 & \cdots & 0  \\
\gamma & 1 - 2 \gamma & \gamma & \ddots & \vdots  \\
0 & \ddots & \ddots & \ddots & 0  \\
\vdots & \ddots & \gamma & 1 - 2 \gamma & \gamma \\
0 & \cdots & 0 & \gamma & 1 - 2 \gamma
\end{pmatrix}
\begin{pmatrix}
w_{1j} \\
w_{2j} \\
\vdots \\
w_{n-2 j}  \\
w_{n-1 j}  \\
\end{pmatrix}
+ \gamma
\begin{pmatrix}
w_{0j} \\ 0  \\ \vdots \\ 0 \\ w_{nj}
\end{pmatrix}
$$

Esto se puede escribir de forma vectorial/matricial como

$$\bmw_{j+1} = A \bmw_j + \bmc \tag{9} $$
donde

$$
A =
\begin{pmatrix}
1 -2 \gamma & \gamma & 0 & \cdots & 0  \\
\gamma & 1 - 2 \gamma & \gamma & \ddots & \vdots  \\
0 & \ddots & \ddots & \ddots & 0  \\
\vdots & \ddots & \gamma & 1 - 2 \gamma & \gamma \\
0 & \cdots & 0 & \gamma & 1 - 2 \gamma
\end{pmatrix}
$$
donde
$$ \bmc =
\gamma \begin{pmatrix}
w_{0j} \\ 0  \\ \vdots \\ 0 \\ w_{nj}
\end{pmatrix} $$
y $\bmw_{j+1}$ es el vector de salida
y $\bmw_{j}$ es el vector de entrada.



## Analisis de estabilidad.
El sistema (9) es iterativo pero necesitamos saber si converge o no. La convergencia de (9) esta supediata al radio espectral de $A$ es decir queremos que $\rho(A) < 1$.

Para este analisis pensemos en los siguientes criteriors.

* Si la solucion fuera analitica no habria error. Pero la solucion es  numerica.

* Los datos en la practica son adquiridos con instrumentacion la cual trae errores de intrumentacion.

* El sistema numerico tambien es impreciso y los errores se propagan.

Definamos el error inicial como

$$\bme_0 = \bmu_0 - \bmw_0  $$
$$\bmw_0 = \bmu_0 - \bme_0  \tag{10} $$
La pregunta es como se propaga este error?
Veamos (de la ecuacion (9) y la (10) )

\begin{eqnarray}
\bmw_1  &=& A \bmw_0 + \bmc = A \bmu_0 - A \bme_0 + \bmc \\
 \bmw_2 &=& A^2 \bmu_0 - A^2 \bme_0 + A \bmc + \bmc  \\
 &\vdots& \\
 \bmw_{j+1} &=& A^{j+1} \bmu_0 - A^{j+1} \bme_0 +
 \sum_{i=0}^j A^i \bmc
\end{eqnarray}
La solucion exacta al comienzo es $\bmu_0$ se propaga hasta
$$A^{j+1} \bmu_0 + \sum_{i=0}^j A^i \bmc $$

La magnitud del termino que me genera el error en este proceso es


$$ \| A^{j+1} \bme_0 \|  $$

Para  el error no crezca queremos que

$$ \| A^{j+1} \bme_0 \|  \le \| \bme_0 \|  $$, despues de algun $j \gg 1$.

Ahora bien, de la norma de matrices

$$\| A^{j+1} \bme_0 \|  \le \| A^j  \| \| \bme_0 \| $$
Dos puntos

1. $\| \bme_0 \|$ no cambia. Es una constante de como los datos ya vienen con ruido.

2. Quremos que $\lim_{j \to \infty} \| A^j \| = 0$.
Es decir, queremos que $\rho(A) < 1$.  (esto es suficiente y se explico en la primera parte del curso.)
