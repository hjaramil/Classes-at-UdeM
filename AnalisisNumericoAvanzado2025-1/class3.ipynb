**Definicion 1.1.7 (Disco de Gershgorin)**: Para una matriz $A \in \mathbb{C}^{n \times n}$ con componentes $a_{ij}$, definimos el radio
$$ R_i = \sum_{j \ne i} | a_{ij}| $$ el centro es $a_{ii}$. El disco
$D(a_{ii}, R) \subset \mathbb{C}$ y radio $R$ se llama el **disco de Gershgorin**.

**Ejemplo 1.1.1**:  Sea $A$ la matriz compleja


$$A = \begin{pmatrix} 3 & \mathrm{i} & 1 \\ -1 & 4 + 5 \mathrm{i} & 2 \\ 2 & 1 & -1  \end{pmatrix}  $$

Los centros y radios fila por fila de la matriz estan dados por

$$\text{fila 1}: \text{centro  } (3,0)  \quad, \quad \text{radio  } |\mathrm{i}| + | 1 | = 2    $$

$$\text{fila 2}: \text{centro  } (4,5)  \quad, \quad \text{radio  } |1| + | 2 | = 3    $$

$$\text{fila 2}: \text{centro  } (-1,0)  \quad, \quad \text{radio  } |2| + | -1 | = 3    $$

**Teorma 1.1.4**(circulo de Geshgorin): Todos los autovalores de una matriz $A \in \mathbb{C}^{n \times n}$ estan en la union de los circulos de Gershgorin.


Prueba: Sea $\lambda$ un autovalor de $A$. Tomemos el autovector correspondiente $\bmx$. Existe alguna componente no 0 (por este autovector no puede ser 0), sea esta $x_j \ne 0$. Sea $i$ tal que $|x_i| \ge | x_j|$,  para $j=1,2, \cdots, n$

Definamos un nuevo vector

$$\bmy = \frac{\bmx}{|x_i|} $$.
Ahora de $A \bmy = \lambda \bmy $. En forma granulada

$$ \sum_j a_{ij} y_j = \lambda y_i $$
pero $y_i=1$.
Esta ecuacion la vamos a separar en dos partes. La diagonal y la no diagonal


$$ \sum_{j \ne i} a_{ij} y_j  + a_{ii}= \lambda y_i $$
$$ \sum_{j \ne i} a_{ij} y_j  + a_{ii}= \lambda $$

Podemos escribir entonces
$$ | \lambda - a_{ii} | = \left | \sum_{j \ne i} a_{ij} y_j  \right |  \le \sum_{j \ne i} |a_{ij} | y_j| \le  \sum_{j \ne i} | a_{ij} | = R_i $$


como $\lambda$ fue arbitrario todos los $\lambda$ estan en la union de todos los circulos.

Volvamos a la recursion inicial



$$\bmx_i = T \bmx_{i-1} + \bmc \tag{1} $$
Si el limite existe entonces $\lim_{i \to \infty} x_i = x$, entonces
en el limite

$$\bmx = T \bmx + \bmc  $$
$$I \bmx = T \bmx + \bmc  $$
Pasamos a restar
$$(I - T)  \bmx + \bmc  $$
$I - T$ es invertible entonces

$$\bmx = (I -T)^{-1} \bmc $$

Para entender como es esta inversa pensemos en la siguiente identidad algebraica


$$ \frac{1 - t^n}{1-t} = 1 + t + t^2 + \cdot + t^{n-1} $$
Esta identidad se puede extender a operadores por que cumplen las leyes asociativas y distributivas de la misma forma


$$ \frac{I - T^n}{I-T} = 1 + T + T^2 + \cdot + T^{n-1} $$

Justifiquemos esta expresion

$$I - T^2 = (I-T)(I+T) $$
$$I - T^3 = (I-T)(I+T + T^2)  $$
$$\vdots  $$
$$I - T^n = (I-T)(I+T + T^2 + \cdots T^{n-1})  \tag{2} $$

Que pasa si la matriz $T$ es convergente. Entoces $\lim_{n \to \infty} T^n=0$


Tomamos el limite a ambos lados cuando $n \infty $ en la ecuacion (2)
$$I  = (I-T)(I+T + T^2 + \cdots T^{n-1} + \cdots  )   $$

De forma que tenemos

$$(I-T)^{-1} = \frac{1}{I - T} = I + T + T^2 + \cdots + T^n + \cdots  $$


**Teorema 1.1.5**: Asuma que la matriz $T$ de la recursion 1 es tal que
$\rho(T)<1$. Esto  equivale qu que la inversa de $(I-T)$ existe y esta dada por

$$(I-T)^{-1} = \frac{1}{I - T} = I + T + T^2 + \cdots + T^n + \cdots = \sum_{i=0}^{\infty} T^i $$

Demostracion
1. $\Rightarrow$: Asumimos que $\lambda$ es un autovalore de $T$ con $\rho(T) < 1$. Es decir para cada autovalore de $T$ tal que


$$T \bmx  = \lambda \bmx$$
$$-T \bmx  = -\lambda \bmx$$
$$\bmx -T \bmx  = \bmx -\lambda \bmx$$
$$I \bmx -T \bmx  = \bmx -\lambda \bmx$$
$$(I  -T )\bmx  = (1 - \lambda ) \bmx$$

En palabras "si $\bmx$ es autovector de $T$ con autovalor $\lambda$, entonces $1-\lambda$ es autovalor de $I-T$ con autovector $\bmx$.


Creamos la siguiete recursion


\begin{eqnarray}
\bmx_i &=& T \bmx_{i-1} + \bmc  \\
\bmx_i &=& T (T \bmx_{i-2} + \bmc) + \bmc  \\
\bmx_i &=& T^2  \bmx_{i-2} + (T+I) \bmc   \\
\bmx_i &=& T^3  \bmx_{i-3} + (T^2 + T+I) \bmc   \\
\vdots  \\
\bmx_i &=& T^i \bmx_{i-i} +  (I + T + T^2 + \cdots + T^{i-1}) \bmc \\
\bmx_i &=& T^i \bmx_0 +  (I + T + T^2 + \cdots + T^{i-1}) \bmc \tag{4}
\end{eqnarray}

Dado que $\rho(T)<1$ tenemos $\lim_{i \to \infty} T^i \bmx_0=0$.
Entonces, tomando el limite al infinito

$$\bmx = \lim_{i \to  \infty} \bmx_i  = (I + T + T^2 + \cdots + T^n + \cdots ) \bmc = (I - T)^{-1} \bmc \tag{5} $$

O sea que la recursion (1) converge y la respuesta esta arriba.

2. $\Leftarrow$: Probaremos que si la serie converge (la inversa existe) entonces $\rho(T)<1$.

Asumamos que el inverso de $I-T$ existe

Dado que $(I-T)^{-1}=I + T + T^2 + \cdots + T^n  + \cdots $, exite, entonces $\lim_{n \infty} T^n = 0$.  Como $T$ converge (por lo que acabo de decir) entonces $\rho(T) < 1$ por teorema de clase anterior.
Fin de teorema.

Un comentario. Existe un caso para el cual $(I-T)^{-1}$ no existe el radio espectral es 1 y el sistema tiene infinitas soluciones.

Sea $T=I$. Entonces

$$I-T=0 $$
para el cual no existe inverso, pero el sistem

$$\bmx_i = T \bmx_{i=1} + \bmc $$
Escoja  $\bmc=0$, y $x$ cualquiera, y $T=I$

$$Ix = x $$
Este caso no esta en el teorema anterior por que aca $\rho(I)=1$ y el teorema anterior necesita que $\rho(T)<1$.

### Analisis de error
Vamos a estudiar el error entre la solucion exacta y las soluciones parciales.

Para esto escribimos $\bmx - \bmx_i$ usando las ecuaciones (4) y (5)

\begin{eqnarray}
\bmx - \bmx_i &=& (\cancel{I + T + \cdots + T^{i-1}} + \cdots + T^{i+k} + \cdots ) \bmc - T^i \bmx_0 - (\cancel{I + T + \cdots + T^{i-1}}) \bmc \\
&=& (T^i + T^{i+1} + \cdots + T^{i+k} + \cdots ) \bmc - T^i \bmx_0 \\
&=& T^i( I + T + \cdots + T^k + \cdots ) \bmc - T^i \bmx_0 \\
&=& T^i \bmx - T^i \bmx_0 \\
&=& T^i (\bmx -  \bmx_0) \\
\end{eqnarray}

Tomando normas vemos que (recuerden que en matrices $ \| A B \| \le \| A \| \| B \|$


$$\| \bmx - \bmx_i \| \le  \| T \|^i \| \bmx - \bmx_0 \| $$

este resultado es muy importante, por que nos dice que la velocidad de convergencia depende de que tan cerca estemos de la solucion $\bmx$ y de que tan pequenha es la norma de $\| T \|$ (la cual debe ser menor que 1).







El problema aca es que no conodemos la solucion exacta $\bmx$, entonces vamos mas bien a comparar con soluciones vecinas.o


\begin{eqnarray}
\| \bmx_{i+1} - \bmx_i  \| &=& \| T \bmx_i + \bmc - (T \bmx_{i-1} + \bmc) \| \\
&=& \| T \bmx_i - T \bmx_{i-1} \| \\
&=& \| T ( \bmx_i - \bmx_{i-1} \| \\
&\le& \| T \|  \| \bmx_i - \bmx_{i-1} \| \\
&\le& \| T \|^2  \| \bmx_{i-1} - \bmx_{i-2} \| \\
&\le& \| T \|^3  \| \bmx_{i-2} - \bmx_{i-3} \| \\
&\vdots& \\
&\le& \| T \|^i \| \bmx_1 - \bmx_0 \|
\end{eqnarray}

Por ahora quiero el error entre $\bmx_i$ y $\bmx_j$.
Asumamos que $i > j \ge 1$, y escribamos


Usamos la propiedad **telescopica**
\begin{eqnarray}
\| \bmx_i - \bmx_j \|  &=& \| \bmx_i - \bmx_{i-1} + \bmx_{i-1}  - \bmx_{i-2} + \bmx_{i-2} \cdots + -\bmx_{j+1} + \bmx_{j+1} - \bmx_j \| \\
&=& \le \| \bmx_i - \bmx_{i-1} \| + \| \bmx_{i-1} - \bmx_{i-2} \| + \cdots + \| \bmx_{j+1} - \bmx_j \| \\
&=&  \left ( \| T \|^{i-1} + \|  T \|^{i-2} + \cdots + \| T \|^j)  \right ) \| \bmx_1 - \bmx_0 \| \\
&=&  \| T \|^j \left ( 1 + \| T \| + \cdots \| T \|^{i-j-1} \right ) \| \bmx_1 - \bmx_0 \| \\
&=& \le \frac{1 - \| T \|^{i-j}}{1 - |\ T \|} \| \bmx_1 - \bmx_0 \|
\end{eqnarray}

Este resultado muestra el tamanho del error entre $\bmx_i$, $\bmx_j$ con respecto al error en la primera iteracion (entre $\bmx_1 y \bmx_0$ y el tamanho de la matriz $T$ .

Escribamos de nuevo este resultado

\begin{eqnarray}
\| \bmx_i - \bmx_j \|  
 \le \frac{1 - \| T \|^{i-j}}{1 - |\ T \|} \| \bmx_1 - \bmx_0 \| \tag{6}
\end{eqnarray}


# Metodos iterativos para resolver sistemas lineales de la forma $A \bmx = \bmb$.

La idea para resolver sistemas lineales es partir la matriz $A$ en tres pedazos

$$A = L + D + U $$
Donde $L$ es triangular inferior  (es decir $L=(a_{ij}$, con $0$ si
$i \le j$. $D$ es diagonal $D=(a_{ii})$ y $U$ es triangular superior, es decir $U = (a_{ij})$ pero con ceros cuando $i \ge j$.



## Metodo de Jacobi
$$A = L + D + U $$
$$D = A - L - U$$

$$A \bmx = \bmb $$
$$(L + D + U) \bmx = \bmb $$
$$ D \bmx = -(L + U) \bmx +  \bmb $$
