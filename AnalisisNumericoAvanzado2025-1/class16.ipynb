# Dimensiones superiores
Veremos en esta parte final del curso la ecuacion de onda en dos dimensiones espaciales $x,y$ y tiempo $t$.

La ecuacion de onda en dos dimensiones espaciales $\mathbf{x}=(x,y)$ y una temporal $t$ es

$$\nabla^2 u(\mathbf{x},t) - \frac{1}{c^2(\mathbf{x})}
\frac{\partial^2 u(\mathbf{x}, t)}{\partial t^2} = -g(\mathbf{x}, t) $$

Donde

$$\nabla^2 u(\mathbf{x}, t) = \Delta  u(\mathbf{x},t) =
\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} $$

Asumimos que $\mathbf{x} = (x,y) \in [a, b ] \times [c, d]$

Podemos asumir condiciones de frontera de Direchlet

* En espacio
$$u(a, y, t) = p(y,t) \quad,  \quad u(b,y,t)=q(y,t) $$
$$u(x, c, t) = r(x,t) \quad,  \quad u(x,d,t)=s(x,t) $$

* En tiempo (iniciales)
$$u(x,y, t_0) = m(x,y)  \quad , \quad u_t(x,y,t_0) = n(x,y) $$

Para evitar confusiones y muchos indices abajo, usaremos el indice de tiempo arriba, como un super-indice
El indice $k$ es para tiempo $i, j$ es para espacio ($(x,y)$)

$$w_{ij}^k = u(x_0 + i \Delta x, y_0 + j \Delta y, t_0 + k \Delta t) $$

Vamos a formular el esquema de diferencias finitas entradas tanto en tiempo como en espacio.

$$ \frac{w_{i+1 j}^k - 2 w_{ij}^k + w_{i-1 j}^k}{\Delta x^2}  - \frac{w_{i j+1}^k - 2 w_{ij} + w_{i j-1}^k}{\Delta y^2} -\frac{1}{c_{ij}} \frac{w_{ij}^{k+1} - 2 w_{ij}^k + w_{ij}^{k-1}}{\Delta t^2} = -g_{ij}^k$$

Debemos despejar $w_{ij}^{k+1}$, para esto multiplicamos todo por $-c_{ij} \Delta t^2$

$$w_{ij}^{k+1} - 2 w_{ij}^k + w_{ij}^{k-1} = \frac{c_{ij}^2 \Delta t^2}{\Delta x^2} (w_{i+1 j}^k - 2 w_{ij}^k + w_{i-1 j}^k) + \frac{c_{ij}^2 \Delta t^2}{\Delta y^2} (w_{i j+1}^k - 2 w_{ij}^k + w_{i  j-1}^k + c_{ij}^2 \Delta t^2 g_{ij}^k $$



Despejando $w_{ij}^{k+1}$
llamando
$$\alpha_{ij} = \frac{c_{ij} \Delta t}{\Delta x} $$
$$\beta_{ij} = \frac{c_{ij} \Delta t}{\Delta y} $$


Entonces



$$w_{ij}^{k+1} =
  2 w_{ij}^k - w_{ij}^{k-1} +  \alpha^2 (w_{i+1 j}^k - 2 w_{ij}^k + w_{i-1 j}^k) + \beta^2 (w_{i j+1}^k - 2 w_{ij}^k + w_{i  j-1}^k) + c_{ij}^2 \Delta t^2 g_{ij}^k  $$

Reordenando

$$w_{ij}^{k+1} =  \alpha^2 (w_{i+1 j}^k + w_{i-1 j}^k)
+ \beta_{ij}^2 ( w_{ij+1}^k + w_{i j-1}^k)
+ 2( 1 - \alpha_{ij}^2 - \beta_{ij}^2) w_{ij}^k - w_{ij}^{k-1} + c_{ij}^2 \Delta t^2 g_{ij}^k  \tag{1} $$

* Analisis de estabilidad
Se asume para el analisis de estabilidad

$$w_{ij}^k = \mathrm{e}^{\mathrm{i}(i k_x \Delta x + j k_y \Delta y + k \omega \Delta t)} $$

El factor de ganancia $G$

$$G = \frac{w_{ij}^{k+1}}{w_{ij}^k}
=
\frac{\mathrm{e}^{\mathrm{i}(i k_x \Delta x + j k_y \Delta y + (k+1) \omega \Delta t)}}{
\mathrm{e}^{\mathrm{i} ( i k_x \Delta x + j k_y \Delta y + k \omega \Delta t )}
} = \mathrm{e}^{\mathrm{i} \omega \Delta t}
$$
Para convergencia se necesita $| G | \le 1$.

La Ecuacion (1) es

$$w_{ij}^{k+1} =  \alpha^2 (w_{i+1 j}^k + w_{i-1 j}^k)
+ \beta_{ij}^2 ( w_{ij+1}^k + w_{i j-1}^k)
+ 2( 1 - \alpha_{ij}^2 - \beta_{ij}^2) w_{ij}^k - w_{ij}^{k-1} + c_{ij}^2 \Delta t^2 g_{ij}^k  $$
Ignormalos la fuente $g=0$
Multiplicamos por $G$.

Voy a llamar $\alpha=\alpha_{ij}$, $\beta=\beta_{ij}$
$$G =  2 \alpha^2 \cos ( k_x \Delta x) + 2 \beta^2 \cos (k_y \Delta y) + 2 (1 - \alpha^2 - \beta^2) -G^{-1}$$


Esto se puede simplificar como

$$G^2 - \phi G + 1 = 0 \tag{2} $$
donde
$$\phi = 2 \alpha^2 \cos (k_x \Delta x) + 2 \beta^2 \cos ( k_y \Delta y) + 2(1 - \alpha^2 - \beta^2) $$

Recuerden

$$\cos \theta = 1 - \sin^2 \frac{\theta}{2} $$
De forma que

$$2 \alpha^2 ( \cos ( k_x \Delta x) -1) = 2 \alpha^2
\left ( 1 - 2 \sin^2 \frac{k_x \Delta x}{2} -1 \right )
=-4 \alpha^2 \sin^2 \left ( \frac{k_x \Delta x}{2} \right ).  $$

De la misma forma para $\beta^2$

$$2 \beta^2 ( \cos ( k_y \Delta y) -1) = 2 \beta^2
\left ( 1 - 2 \sin^2 \frac{k_y \Delta y}{2} -1 \right )
=-4 \beta^2 \sin^2 \left ( \frac{k_y \Delta y}{2} \right ).  $$






De esta forma queda
$$\phi =  -4 \alpha^2 \sin \left ( \frac{k_x \Delta x}{2}  \right ) - 4 \beta^2 \left ( \frac{k_y \Delta y}{2} \right ) + 2 = 2 \psi $$

Donde

$$\psi = -2 \alpha^2 \sin^2(k_x \Delta x/2) - 2 \beta^2 \sin^2 (k_y \Delta y/2) + 1 $$

La solucion de la cuadratica

$$G = \frac{\phi \pm \sqrt{\phi^2 - 4}}{2} =\frac{2 \psi \pm \sqrt{4 \psi^2 - 4}}{2} = \psi \pm \sqrt{\psi^2 -1}$$

Casos. Si $|\psi | > 1 $, entonces diverge por que, tomando el "+" $|\psi + \sqrt{\psi^2 - 1}| > 1$.
Ahora si $\psi < -  1$, entonces tomando el "-"

$$\psi - \sqrt{\psi^2 -1 } < - 1 $$

La unica opcion que puede funcionar es $| \psi | \le 1$.

En este caso podemos escibir $G$

$$G = \psi \pm \mathrm{i} \sqrt{1 - \psi^2}  $$

En este caso

$$| G |^2 = \psi^2 + (1 - \psi^2) = 1 $$
Lo que hace que el sistema sea estable.

Que significa $| \psi | \le 1$?

$$|-2 \sin^2(k_x \Delta x/2) - 2 \beta^2 \sin^2 (k_y \Delta y/2) + 1 | \le 1 $$

Es decir
Recuerden

$$| x | \le a  \Leftrightarrow -a \le x \le a$$


$$-1 \le -2 \sin^2(k_x \Delta x/2) - 2 \beta^2 \sin^2 (k_y \Delta y/2) + 1  \le 1 $$


restando 1
\begin{eqnarray*}
  -2  \le  -2 \alpha^2 \sin^2(k_x \Delta x/2) -2 \beta^2 \sin^2 (k_y \Delta y/2)  \le 0,
\end{eqnarray*}
Es decir
\begin{eqnarray*}
  2  \ge  2 \alpha^2 \sin^2(k_x \Delta x/2) +  2 \beta^2 \sin^2 (k_y \Delta y/2)
\end{eqnarray*}
o, de otra forma

\begin{eqnarray*}
  \alpha^2 \sin^2 ( k_x \Delta x/2) + \beta^2 \sin^2 (k_y \Delta y/2) \le 1
\end{eqnarray*}

Podemos asegurar esta desigualdad si $\alpha \le 1/\sqrt{2}$ y $\beta \le 1/\sqrt{2}$.
Como $\alpha=c \Delta t/\Delta x$ y $\beta=c \Delta t/ \Delta y$, entonces
debemos tener que


\begin{eqnarray*}
  \Delta t^2 \left ( \frac{c^2}{\Delta x^2} + \frac{c^2}{\Delta y^2} \right ) \le 1
\end{eqnarray*}
de donde, la condicion para $\Delta t$ se obtiene despejando $\Delta t$,
es decir,


\begin{eqnarray*}
  \Delta t \le \frac{1}{c \sqrt{\frac{1}{\Delta x^2} + \frac{1}{\Delta y^2}}}
\end{eqnarray*}
En una dimensión (por ejemplo $x$ solamente) podemos pensar que $\Delta y=0$ y tenemos que
\begin{eqnarray*}
  \Delta t \le \frac{1}{c \Delta x}
\end{eqnarray*}
Que coincide con la f'ormula para el caso de una dimension espacial.


Si la velocidad es variable podemos buscar una cota para $\Delta t$.
Es decir,
\begin{eqnarray*}
  \Delta t \le \max_{c_{ij}} \frac{1}{c_{ij}
  \sqrt{\frac{1}{\Delta x^2} + \frac{1}{\Delta y^2}}} .
\end{eqnarray*}
Si comparamos esta ecuaci'on con la ecuaci'on de estabilidad para una dimension espacial vemos que
esta es una generalizaci'on de la condici'on de estabilidad para el caso
de una dimensi'on, con una restricci'on mayor puesto que la
desigualdad debe cubrir las dos direcciones $x$ y $y$.
