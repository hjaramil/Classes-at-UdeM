## Matriz adjunta.
La conjugada de la matriz $A$ se nota como $B= \overline{A}$. Donde $b_{ij} = \overline{a_{ij}}$.
El conjugado de un numero complejo $a  + \mathrm{i} b$ es $a - \mathrm{i} b$, donde $a,b \mathbb{R}$. 

Por ejemplo

$$A = \begin{pmatrix} 3 -2 \mathrm{i}  & 1 + \mathrm{i} \\ 1  & -2 - \mathrm{i} \end{pmatrix} $$

$$\overline{A} = \begin{pmatrix} 3 +2 \mathrm{i}  & 1 - \mathrm{i} \\ 1  & -2 + \mathrm{i} \end{pmatrix} $$

La matriz **adjunta** es la transpuesta de la conjugada. La notamos como $A^*$. Es decir

$$A^* = (\overline{A})^T = \overline{A^T} $$

En terminos de componentes si $B=A^*$,
$b_{ij} = \overline{a_{ji}}$.

Ejemplo:

$$A^* = \begin{pmatrix} 3 +2 \mathrm{i}  & 1 \\ 1 - \mathrm{i}    & -2 + \mathrm{i} \end{pmatrix} $$

Nota: Hay muchos casos donde las matrices son ortogonales y la inversa de una matriz ortogonal es la matriz adjunta. 


Definicion [producto interno]: Sean $u, v \in \mathbb{C}^n$. entonces defininmos el **producto interno** de $u$ con $v$ como $u^* v$. Es decir, si 
$u=(u_1, \cdots, u_n)^T$, $v=(v_1, \cdots, v_n)^T$. 

$$c = \langle u , v \rangle = u^* v =  \sum_{i=1}^n \overline{u}_i v_i$$

Definicion (ortogonalidad). Dos vectores $u,v$ son ortogonales si su producto interno es 0. Es decir si

Existen tambien matrices ortogonales. La matriz $A$ es ortogonal si sus columnas son dos a dos ortogonales.
Es decir si $A=[A_1 | A_2 | \cdots | A_n ]$ entonces $A$ es ortogonal si $\langle A_i, A_j \rangle = 0$ para $i \ne j$. 

Un ejemplo: La matriz de rotacion:

$$ A = \begin{pmatrix}  \cos \theta & \sin \theta \\

$A_1 = (\cos \theta, - \sin \theta)^T$, $A_2=(\sin \theta, \cos \theta)^T$. 
$$\langle A_1, A_2 \rangle = \cos \theta \sin \theta - \sin \theta \cos \theta = 0 $$

La "longitud" de un vector. Sea $u \in \mathbb{C}^n"$, definimos la **norma** de $u$ como

$$ \| u \| = \sqrt{ \langle u , u \rangle}  
=\sqrt{ u^* u}  = \sqrt{ \sum_{i=1}^n \overline{u}_i u_i}$$ donde $u=(u_1, \cdots, u_n)^T$ .

Si vector tiene longitud 1 se dice que es **unitario**. 
Es decir $\| u \|=1$ entonces $u$ es unitario.

Si las columna de una matriz ortogonal son unitarias, entonces la matriz es **ortonormal**. Ejemplo.

$$ A = \begin{pmatrix}  \cos \theta & \sin \theta \\
-\sin \theta & \cos \theta \end{pmatrix} $$.

$$ \| A_1 \|= \sqrt{ \langle A_1, A_1 \rangle}= \sqrt{
( \cos \theta, - \sin \theta) \begin{pmatrix} \cos \theta \\ -\sin \theta \end{pmatrix} } = \sqrt{\cos^2 \theta + \sin^2 \theta} = 1$$. O sea que $A_1$ es unitario. Se deja como prueba verificar que $A_2$ tambien es unitario.


**Ojo** en la literatura se dice que una matriz es ortogonal cuando es ortonormal. 

$$ B = \begin{pmatrix}  2 \cos \theta & \sin \theta \\
-2 \sin \theta & \cos \theta \end{pmatrix} $$.

$B_1 = 2 ( \cos \theta, - \sin \theta)^T$, 
$B_2 = (\sin \theta, \cos \theta)^T$.

$$\langle B_1, B_2 \rangle = 2 \cos \theta \sin \theta - 2 \sin \theta \cos \theta = 0  $$.
$\| B_2 \| = 2$. La primera columna no es unitaria. 
Segun esto , para literatura $B$ no es ortogonal.

### Autovalores y autovectores.
Sea $A$ una matriz $n \times n$ y sea $x \in \mathbb{C}^n$ , $x \ne 0$ y $\alpha \in \mathbb{C}$ tales que

$$ Ax = \lambda x $$
Entonces $x$ se llama un **autovector** de $A$ y
$\lambda$ se llama un **auttovalor** de $A$.
Tambien se llaman **vectorres caracteristicos**
y **valores caracteristicos**

[Wikipedia](https://en.wikipedia.org/wiki/Eigenvalues_and_eigenvectors)  ofrece unas graficas con o sin animacion muy ilustrativas del concepto.



Como se calculan los autovectores y autovalores.

Recuerde que $x = Ix$.

\begin{eqnarray}
A x &=& \lambda x \\
A x - \lambda x &=& 0 \\
A x - \lambda Ix &=& 0 \\
(Ax - \lambda I)x &=& 0 
\end{eqnarray}
Un teorema del algebra lineal dice que si $Bx=0$ y $x \ne 0$ entonces el determinante de $B$ es 0. 

O sea que

$$ p(\lambda) = \det ( Ax - \lambda I) = 0$$

Ej:
\begin{eqnarray}
A = \begin{pmatrix} 3 & 6 \\
-1 & 2
\end{pmatrix}
\end{eqnarray}

\begin{eqnarray}
A - \lambda I = \begin{pmatrix} 3 -\lambda  & 6 \\
-1 & 2 - \lambda
\end{pmatrix}
\end{eqnarray}





\begin{eqnarray}
\det (A - \lambda I) =  \det \begin{pmatrix} 3 -\lambda  & 6 \\
-1 & 2 - \lambda
\end{pmatrix} = (3-\lambda)(2-\lambda) .
\end{eqnarray}
