Teorema: Si $A$ es autoadjunta, los autovectores correspondientes a autovalores distintos son ortogonales y los autovalores son reales. 

**Prueba**: Asumamos, sin perdida de generalidad, que $\| u_j \|=1$, donde $u$ es autovector y $\lambda_j$ el autovalor correspondiente. Usamos la forma cuadratica $\langle A u_j, u_j \rangle $.


$$\langle A u_j, u_j \rangle = \langle \lambda_j u_j, u_j \rangle =   \overline{\lambda_j} \langle u_j, u_j \rangle = \overline{\lambda_j }  \quad (1)  $$
Ahora bien, como $A=A^*$, entonces:

$$\langle A x, y \rangle = (Ax)^* y = x^* A^* y =x^*(A^* y) = \langle x, A^* y \rangle  \quad (2) $$


Usamos esta propiedad:
empezamos con el final de la ecuacion (1)

$$ \overline{\lambda_j} = \langle A u_j, u_j \rangle = \langle u_j, A^* u_j \rangle =\langle u_j, A u_j \rangle = \langle u_j, \lambda_j u \rangle = \lambda_j \langle u_j, u_j \rangle = \lambda_j  $$
o sea que
$\overline{\lambda_j} = \lambda_j$ o sea que $\lambda_j$ es real. 

Probemos ahora que autovectores correspondientes a atuvalores distintos son ortogonales.

Sean $\lambda_1 \ne \lambda_2$ autovalores de los autvectores $u,v$.

Calculemos $\langle Au, v \rangle $

$$ \langle Au, v \rangle = \langle \lambda_1 u, v \rangle = \lambda_1 \langle u, v \rangle   \quad (3) $$
como $A^* = A$,

$$ \langle Au, v \rangle = \langle u , A v \rangle = \langle u, \lambda_2 v \rangle = \lambda_2 \langle u, v \rangle  \quad (4)  $$
donde usamos la propiedad (2).

La (3) es igual a la (4).  Es decir
$$ \lambda_1 \langle u , v \rangle = \lambda_2 \langle u, v \rangle $$.
Pasando a restar

$$(\lambda_1 - \lambda_2) \langle u, v \rangle = 0 $$
pero como $\lambda_1 \ne \lambda_2$ , $\lambda_1 - \lambda_2 \ne 0$, entonces

$$ \langle u, v \rangle = 0$$.

Asuma que una matriz $A$ tiene todos sus autovalores distintos de par en par. Entonces todos sus autvectores son ortogonales. En este caso los podemos **normalizar**
y forman una base ortonormal, que la podemos poner columna por columna en una matriz $Q$. Sean vectores los $x_i$. Como son normalizados $x_i^* x_j = \delta_{ij}$. 

De esta forma construimos para $i,j=1,2, \cdots, n$
construimos

$$x_i^* A x_j = x_i^* \lambda_j x_j = \lambda_j x_i^* x_j = \lambda_j \delta_{ij} = \lambda_i $$

Interpretemos:
* La matriz $Q=[x_1 | x_2 | \cdots | x_n]$, es la matriz de autovectores de $A$.
* La matriz $\Lambda$ es una matriz diagonal donde la las entradas son $\lambda_j \delta_{ij}$, es decir los autovalores de $A$ correspondientes a los $x_i$.

Obtenemos:

$$ \Lambda = Q^* A Q  \quad (5) $$.
Recordemos: $\Lambda$ es una matriz diagonal $n \times n$ con los autovalores $\lambda_i$ de los autovectores $x_i$ que forman las columnas de $Q$.


Si en la ecuacion (5) multiplicamos a izquierda por $Q$ y a derecha por $Q^*$ (recuerde que $Q^* Q = I = Q Q^*$.)


$$Q \Lambda Q^* = Q Q^* A Q^* Q = I A I = A $$
Encontramos

$$A = Q \Lambda Q^* $$


En palabras: Si $A$ es autoadjunta, $A$ se puede escribir (factorizar) como el triple producto de la matriz $Q$ de
autovectores de $A$ con la matriz diagonal $\Lambda$ de autovalores de $A$ (en correspondencia) y la conjugada de $Q$, es decir, $Q^*$.  A esto se le llama ** el teorema de descomposicion espectral con autovalores distintos**.

Un espacio lineal es el generado de una base.
$\langle x_1, x_2, \cdots, x_n \rangle = \sum_{i=1}^n \alpha_i x_i  $, donde $\alpha_i \in \mathbb{C}$. 

Defincion [espacio invariante]. Sea $W$ un espacio lineal de dimension $n$. Decimos que $W$ es **invariante** con respecto a $A$ si $AW \subset W$. $A \in \mathbb{C}^{n \times n}$, 

Ejemplos:

* El trivial $W = \{ 0 \}$. Es invariante por que
$AW = \{0\} \subset W$. 
* El trivial $W = \mathbb{C}^n$. 

Probemos que $AW \subset W$. 

Asumamos $x \in AW$, probemos que $x \in W$.

$$ x \in AW $$, es decir $x = Ay$, $y \in \mathbb{C}^n$
pero $A \in \mathbb{C}^{n}$ entonces $x$ es un vector en $\mathbb{C}^n = W$, es decir $x \in W$. 

Estos son los espacio invariantes triviales. Queremos algo mas interesante. 

* El espacio nulo de $A$ es invariante. 

Prueba. Sea $W = \mathcal{N}(A)$. Vamos a probar que $AW \subset W$. 

Asumimos $x \in AW$ y probaremos que $x \in W$.

Si $x \in AW$, $x=0$. Pero $0 \in \mathcal{N}(A)$, es decir
 $x \in W$. Listo. 

 * Asuma un conjunto de autvectores de $A$, $\{x_1, x_2, x_k}$, entonces el gene
 rado de este conjunto es invariante de $A$.

 En simbolos

 $$W = \langle x_1, x_2, \cdots, x_k \rangle =\left  \{ \sum_{i=1}^k \alpha_i x_i | \alpha_i \in \mathbb{C} \right \} $$ (tarea).
* Si $x_1$ es un autovector de $A$ entonces su generado ( $\alpha x$, $\alpha \in \mathbb{C}$ es invariante.

Prueba:

$$ W= \langle x \rangle = \{ \alpha x  | \alpha \mathbb{C} \}$$.

Debemos probar que $AW \subset W$.
Sea $x \in AW$, debemos mostrar que $x \in W$. 
Si $y \in AW$, $y  =  A u$ con $u \in W$, o sea que $

$$ y = A (\alpha x)=  \alpha A x = \alpha \lambda x = \beta x$$
donde $\beta \in \mathbb{C}$.  Es decir que $y \in W$ .
Esto finaliza la prueba.

Lema: Para cada  espacio invariante $W \ne \{0\}$ de una matriz cuadrada $A$, existe un auto-par (autovalor/vector). (eigenpair: engivalue/vector).

Prueba: Asumamos que la dimension de $W$ es $k$, es decir,
existe una base $\{ x_1, x_2, \cdots, x_k \}$ para $W$.
Sabemos que, por el teorema fundamental del algebra, cada matriz $A$ tiene por lo menos un autovalor y un autovector (el teorema dice: cualquier polinomio de orden $n \ge 1$ tiene al menos una raiz en los complejos. En particular el polinomio caracteristico). 

Tenemos como hipotesis $AW \subset W$. Para cada $x \in W$, $Ax \in W$. $x=\sum_{i=1}^k \alpha_i x_i$ con 
$A x_i = \sum_{j=1}^k \beta_{ji} x_j $. Queremos resolver la ecuacion $Ax - \lambda x = 0$ ($Ax = \lambda x $). 

Lo que queremos es que exista $\lambda, x$ autopar tal que

$$ Ax = A x=A \sum_{i=1}^k \alpha_i x_i = \sum_{i=1}^k \alpha_i A x_i = \lambda \sum_{i=1}^k \alpha_i x_i $$

Pasando a restar a la izquierda y reemplazando la expresion con $\beta_{ji}$ de arriba.
\begin{eqnarray}
\sum_{i=1}^k \alpha_i \left ( \sum_{j=1}^k \beta_{ji} x_j - \lambda x_i
    \right ) &=& 0 \\
     \sum_{i=1}^k \alpha_i \left ( \sum_{j=1}^k \beta_{ji} x_j - \lambda \delta_{ij} x_j
    \right ) &=& 0  \\
    \sum_{j=1}^k \left ( \sum_{i=1}^k \alpha_j( \beta_{ji} - \lambda \delta_{ij})  \right ) x_j
\end{eqnarray}

Como los $x_j$ son linealmente independientes.
Los coeficientes en la suma deben ser 0

$$ \sum_{i=1}^k \alpha_j( \beta_{ji} - \lambda \delta_{ij})  = 0$$

Esto en notacion matricial es  $(B - \lambda I) \alpha = 0$. Es decir que $\lambda$ es autovalor de $B$ que siempre existira y la ecuacion es valida.
