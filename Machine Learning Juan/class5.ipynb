**Definition:**[condition number] El numero de condicion de una matriz $A$ es

$$\kappa = \frac{\sigma_1}{\sigma_n} $$
Claramente si $\sigma_n = 0$, el numero de condicion es infinito y la matriz $A$ no tiene inversa. El numero de condicion "ideal" es 1.
Este es el numero para las matrices ortogonales. Si una matriz es ortogonal los valores singulares son 1. (tarea)

Segun Cash Yuste si $\kappa = 10^k$ se pierden $k$ cifras de precision (heuristico). 


Si $\kappa \gg 1$, entonces la matriz $A$ se dice que es "ill conditioned" (mal condicionada) y al sistema $Ax=b$ se dice que es mal condicionado.
$ \kappa \approx 1$ entonces el sistema es "well conditioned" , bien condicionado. "ill condinioned" es inestable. Es decir cambios pequenos en $b$ generan cambios grandes en $x$. 

**Definicion** [Cociente de Rayleigh]  Asuma $A$ es autoadjunta, el cociente de Rayleigh (Rayleigh Quotient) para la matriz $A$ con respecto a $x \in \mathbb{R}^n$ se define como

$$ R(A, x) =  \frac{x^* A x}{x^* x} = \frac{\langle x , Ax \rangle}{\| x \|^2} = \frac{\langle Ax, x \rangle }{\| x \|^2}$$
Tambien podemos escribir:

$$ R(A, x) = \frac{x^*}{\| x^* \|} A \frac{x}{\| x\|}  = u^* A u $$

**Teorema** (Rayleigh): El autovalor mas grande de una matriz autoadjunta $A$ en $\mathbb{C}^{n \times n}$ corresponde al mayor valor del cociente de Rayleigh. 

Sea $\{ u_1, u_2, \cdots, u_n \}$ una base ortonormal del espacio $\mathbb{C}^n$ donde los $u_i$ son los autovectores de $A$. Esto es posible por el teorema espectral. Por algebra lineal I, cada vector $u$ , unitario se puede escribir

$$ u = \sum_{i=1}^n \alpha_i u_i   \quad \alpha_i \in \mathbb{C}$$
como $\| u \|=1$, $\sum | \alpha_i |^2  = 1$.

$$R(A, u) = u^* A u = \sum_{i=1}^n \sum_{j=1}^n \overline{\alpha_i} \alpha_j  u_i^* A u_j $$

Ahora

$$  u_i^* A u_j = u_i^* ( \lambda_j u_j) = \lambda_j u_i^* u_j = \lambda_j \delta_{ij} $$
. 


$$R(A, u) = u^* A u = \sum_{i=1}^n \sum_{j=1}^n \overline{\alpha_i} \alpha_j  (\lambda_j \delta_{ij} ) = \sum_{i=1}^n | \alpha_i |^2 \lambda_i $$
Sea $\lambda_1$ el mas grande de todos los autovalores.



$$ R(A, u) \le \lambda_1 \sum_{i=1}^n | \alpha_i|^2 $$


Como $\sum_{i=1}^n | \alpha_i |^2 = 1$. Yo puedo decir que

$$ R(A, u) \le \lambda_1 $$.
O sea que $\lambda_1$ es una cota superior para el cociente de Rayleigh.

Esa cota se adquire i.e. la cota es parte del conjunto, es decir que existe un $u$ tal que $R(A, u) = \lambda_1$ . Cuando? $\alpha_1=1$, $\alpha_i=0$, $i=2,3, \cdots ,n$ O sea que $u_1$ es el autovector del autovalor $\lambda_1$ mas grande de todos y es igual a $R(A, u_1)$.

El gradiente de la matriz $A$ esta definido por

$$ \nabla_k A = \frac{\partial A}{\partial x_k} $$


**Teorema**: Si $A$ es una matrix real $m \times n$ y $x \in \mathbb{R}^n$ entonces

$$ \nabla (A x) = A $$
donde el gradiente es respecto $x$ y mas exactamente

$$ \frac{ \partial A x}{\partial x_k } = \nabla_k (Ax) = A_k \quad (0) $$
donde $A_k$ es la columna $k$ de la matriz $A$. $A$ se asume constante.

*Prueba*: Por componentes. Tomemos la componente $i$ de $Ax$.

$$ y_i = \sum_{j=1}^n a_{ij} x_j $$

Entonces

$$ \frac{\partial y_i}{\partial x_k} = \sum_{j=1}^n a_{ij} \frac{\partial x_j}{\partial x_k}  = \sum_{j=1}^n a_{ij} \delta_{jk} = a_{ik}  $$

Pero esto para $i=1,2, \cdots, m$ es la columna $k$ de la matriz $A$. Es decir $A_k$. 

**Teorerma**: Asuma $f(X)$, $g(x)$ funciones vectoriales de $x \in \mathbb{R}^n$. Entonces

$$  \frac{\partial \langle f, g \rangle }{\partial x_i} = 
\left \langle \frac{\partial f}{\partial x_i} , g  \right \rangle + \left \langle f , \frac{\partial g }{\partial x_i} \right \rangle  $$

*Prueba:*  Expandemos el producto interno. 

$$ \langle f, g \rangle = \sum_{j=1}^n \overline{f_j(x)} g_j(x)  $$

Tomamos la derivada:

\begin{eqnarray} \frac{\partial \langle f, g \rangle }{\partial x_i} &=& \sum_{j=1}^n  \frac{\partial}{\partial x_i} \overline{f_j(x)} g_j(x) \\
&=& \sum_{j=1}^n \frac{\partial \overline{f_j(x)}}{\partial x_i} g_j(x) + \overline{f_j(x)} \frac{\partial g_j(x)}{\partial x_i}  \\
&=& \sum_{j=1}^n \frac{\partial \overline{f_j(x)}}{\partial x_i} g_j(x) + 
\sum_{j=1}^n \overline{f_j(x)} \frac{\partial g_j(x)}{\partial x_i}  \\
&=& \left \langle \frac{\partial f}{\partial x_i} , g  \right \rangle + \left \langle f , \frac{\partial g }{\partial x_i} \right \rangle
 \end{eqnarray}

 En notacion de gradiente podemos decir

### Minimos cuadrados regularizados.
Sea $A$ una matriz real de orden $m \times n$. $x \in \mathbb{R}^n$. Podemos escribir $A^T$ pero uso el simbolo $A^*$.  Si $A$ tiene inversa
y queremos resolver $A x = b$.
$$ x = A^{-1} b $$
sin embargo esto casi nunca sucede. De hecho casi todas las matrices no son cuadradas.

Pueden producirse dos situaciones.

1. La matriz $A$ esta sobredeterminada. Es decir el sistema $Ax = b$ tiene mas ecuaciones independientes que incognitas. Lo que pasa es que el espacio nulo $\mathcal{A} = \{ 0 \}$ pero el rango $\mathcal{A}$ no cubre todo el espacio. 

2. La matriz $A$ es subdeterminada. Es decir el sistema $Ax = b$ tiene mas incognitas que ecuaciones independientes. Tiene un numero infinoto de solucioness. 

Como resolvemos este problema?  
[Tikhonov](https://en.wikipedia.org/wiki/Tikhonov_regularization) (y otros)
se inventaron una solucion. Poner una funcion "penalti". Es decir ellos
sugieren minimizar 

$$ f(x) = \| A x - b \|^2 + \| \Gamma x \|^2  $$
donde $\Gamma$ es una matriz positiva definidad.  Se resuelve este problema
(tarea para mas adelante). Es decir $\nabla f(x) = 0$, obtenemos:

$$ (A^*A + \Gamma^* \Gamma) x = A^* b  $$

Un caso particular importante de $\Gamma$ es $\Gamma= \sqrt{\lambda} I$, donde $\lambda > 0 $. En este caso

$$ \Gamma^* \Gamma = \lambda I$$
De forma que tenemos:


$$ (A^*A + \lambda I) x = A^* b  $$
De aca sale

$$ x = (A^*A + \lambda I)^{-1} A^* b $$ 

Esta solucion es la minimizacion de la funcion objetivo (de costo)

$$ J(x) = \frac12 \| Ax - b \|^2 + \lambda \| x \|^2 $$
