## Comando SED (Stream Editor)
El SED permite cambiar archivos sin necesitad de entrar con un editor.

Hay tres formas de cambio

* Insertar
* Reemplazar
* Borrar (remover)

La sintaxs

```sed OPTIONS ... [SCRIPT][INPUT FILE]

Ejmplos la "shell" en clase.

Actividades:
* se agregan lineas con ```a```
* agregar lineas de acuerdo a un patron
* Cambiar lineas con patrones
* sustituir una sola linea


Nota: El backslash ```/``` no es necesario, se reemplazar por

* espacios vacios
* puntos
* pipies ```|```

Trate con regexp (regular expressions).

## AWK:

AWK es in lenguaje de programacion en Unix,
El nombre viene de las 3 personas que lo inventaron.
Alfred Aho, Peter Weinberger, Brian Kernighan.

Algunas opciones y propiedades importantes de AWK son:
(solo para archivos de texto).

* Revisa linea por linea
* Entiende un archivo como una matriz.
* Compara lineas con patrones
* Ejecuta acciones de acuerdo a patrones
