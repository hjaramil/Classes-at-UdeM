## Continued Fracions (Fracciones continuadas).

<font size ="12">
$$a_0 + \frac{x}{ a_1 + \frac{x}{a_2 + \frac{x}{a_3 + \frac{x}{a_4 + \frac{x}{a_5+ \frac{x}{a_6}}}}}} .$$
</font>

```python
from sympy import symbols
a = symbols('a0:6')
a
```

```python
x= symbols('x')

def continued_fraction(depth):
    frac = a[0]
    for d in range(0, depth):
        frac = frac.replace(a[d] , a[d]+ x/a[d+1])
    return frac
```

```python
cf = continued_fraction(5)
cf
```

# conversion de ecuaciones a LaTeX
