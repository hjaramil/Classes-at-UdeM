### Discussion about the flag ```name``` in the command ```find```.
En mi curso ```OOP```, trate estos dos comandos.

```
find . -name Felix -type d
find . Felix -type d
```

Olvidar la opcion ```-name``` hace que liste mar archivos de los que queremos.

Se usa ```2> salida``` para archivar erroes en comando ```find```.

Otra cosa que se puede usar es ```2 > /dev/null ```
