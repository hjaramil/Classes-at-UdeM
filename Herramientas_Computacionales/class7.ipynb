Hacemos una lista de las opciones de AWK

* Los simbolos $1, $12, etc, indican el campo (columna). El simbolo $0 es **toda la linea**

* ```NR```: Numero de registro (fila) Veremos ejemplos de este.

* ```NF```: Numero de campo (number field). Numero de columnas, ```$NF ```la ultima columna

* ```FS```: Field Separator (separador de campos, columnas). Por defecto es espacio en blanco, pero pude ser que esten seprados por comas, barras verticales , etc

* ```RS```: Separador de lineas (filas). Los registros  por defecto estan separadas por el ```\n```.  Pero puden estar separadas tambien por vairas lineas en blanco. Por ejemplo  ```\n\n``` es doble linea.

* Existen otras opciones e.g. ```OFS``` , Output field separator
```ORD``` salida de registro. Dejamos esto al estudiante[link text](https://).
