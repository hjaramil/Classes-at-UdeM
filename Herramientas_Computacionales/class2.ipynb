### Actividad \#1

Se discutio acerca del sistema de archivos.

Abrimos una pantalla ```Ctrl-Alt-T``` o con ```Actividades```.

El primer comando para explorar el sistema
es
```
man man
```

Cada vez que se corre un comando la informacion que se tiene es:

* Name: Nombre del comando
* Synopsis: las opcoines del comando rapidas (la descripcion rapida).

* Description: descripcion detallada
* Examples: Ejemplos
* Defaults : los parametros por defecto
* Options: opciones.

Otra de buscar informacion sobre un comando es con
```
info comando
```

El comando ```ls ``` tiene opciones intersantes que vamos a ver mas adelante.

Formas de crear archivos en modo comando.

* Usando ```touch```
* Usando ```echo```
* Usando un editor (vi, emacs, gvim, gedit, Microsoft Word, etc)

El comando ```pwd``` (path word) indica donde estamos parados.

El comando ```whoami``` indica quien es el usuario de esta cuenta.

Para crear el primer archivo que sea un archivo vacio usamos


```
touch nada
````

Una vez creado un archivo, se puede agregar, desde el modo comando informacion usando el simbolo

```>>```

Cambio de privilegios sobre un archivo. Comando

```chmod```


Los permisos se pueden resumir en un numero XYZ,
donde X, Y, Z son octales.

Ejemplo
```chmod 700 leon```

El numero 7 octal es $111$. El usuario puede leer, escribir y ejecutar

```chmod 750 leon```

El 5 binario $101$
El usuario y puede hacer todo, el grupo pude leer y ejecutar, los demas nada.
