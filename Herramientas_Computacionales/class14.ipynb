## Solucion de ecuaciones simultaneas (lineales o no)

Ejemplo:

\begin{eqnarray}
x = 3 y - z \\
5x - 3 z = y
\end{eqnarray}



```python
from sympy import Eq, linsolve
x,y,z = symbols( ' x  y  z')
eq1 = Eq(x, 3*y - z) # x = 3*y - z
eq2 = Eq(5*x - 3*z, y) # 5*x-3*z = y
linsolve([eq1, eq2], x,y,z)
```

```python
# extraer lhs , rhs
display(eq1.rhs)
display(eq2.rhs)
```

Ecuaciones simultaneas con solucion no unica o que no existe


\begin{eqnarray}
2x + 3 y = 5 \\
4x - 2 y = -6 \\
\end{eqnarray}

```python
# definiamos las ecuaciones
from sympy import symbols, Eq, solve
x,y = symbols('x y ')

eq1 = Eq( 2*x + 3*y, 5)
eq2 = Eq( 4*x - 2*y, -6)

solution = solve( (eq1, eq2), (x,y))
print("Solution:")
print(f"x = {solution[x]}, y ={solution[y]} ")
```

## Sistemas no lineales.

Por ejemplo

\begin{eqnarray}
x^2 + y^2 &=& 25 \\
x-y &=& 3
\end{eqnarray}

```python
# simbolos
x,y = symbols('x y')

# ecuaciones
eq1 = Eq(x**2 + y**2, 25)
eq2 = Eq(x-y, 3)


# resuelva
solution = solve( (eq1, eq2), (x,y))

print("Solution")
#print(f"x = {solution[x]}, y ={solution[y]} ") # no es diccionario
solution
```

```python
# graficamos
[x,y] = solution
p = plot_implicit( Or(eq1, eq2), xlabel='x', ylabel='y' , \
                  title='ecuaciones lineales y su solucion', \
                   markers=[{'args':[x,y,'ro']}])
```

## Funciones racionales

```python
x = symbols('x')
f = x**2 + 1
g = x**2 -1

h = 1/f + 1/g
h
```

```python
# simplificar
fs = simplify(h)
fs
```

```python
## Fracciones parciales "apart"
from sympy import apart
fp = apart(fs)
fp
```

```python
fp.args
```

```python
fp.args[0]
```

```python
fp0 = fp.args[0]
fp1 = fp.args[1]
fp2 = fp.args[2]
display(fp0+fp1+fp2)
simplify(fp0+fp1+fp2)
```

```python
# together (juntar)
toghe = (fp0 + fp1 + fp2).together()
toghe
```

```python
simplify(toghe)
```

```python
## sumuliplicar
fp0*fp1
```

```python
# dividir
fp0/fp1
```

```python
# fracciones parciales
apart(fp0/fp1)
```

```python
## Graficar
toPlot = fp0*fp1/fp2
toPlot
```

```python
from sympy import plot
plot( toPlot, x**3, ylim=[-7,7])
```
