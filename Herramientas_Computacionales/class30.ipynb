# Colocar ejes en graficas
Ver notas de clase en el fuente `TikZ`


  *``scale``: escala el nodo
  * ``rotate``: rota el nodo
  * ``below``: lo pone debajo
  * ``above``: lo pone encima
  * ``left``: lo pone a la izquierda
  * ``right``: lo pone a la derecha
  * ``below left``: izquierda abajo (suroeste)
  * ``below right``: izquierda abajo (sureste)
  * ``above right``: izquierda abajo (noreste)
  * ``above left``: izquierda abajo (noroeste)
  * ``xshift``: corre unidades en el eje x
  * ``yshift``: corre unidades en el eje x
