# History
Este comando muestra la historia de los comandos ejecutados.
Cuantos? , el proximo comando lo dice.

` echo $HISTSIZE`

Como se usa el comando history para editar comandos anteriores

- Usar las flechas, iz, der, arri, ab
- `!num`  donde `num` es un numero de la historia
- Cuando escribe `!` y algunas letras, y <enter> ejecuta
el ultimo comando que comienza con esas letras.
- usar el "caret" `^`. Por ejemplo `^old^new` reemplaza
la palabra `old` por la palabra `new` en el comando anterior.
- El comando `!!` ejecuta el ultimo comando que ejecuto
