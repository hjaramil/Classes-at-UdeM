## Linux:

* Es un sistema operativo. Lo vamos a aprender comandos y creamos shell scripts. Programas en lenguaje bash.

* Origen: MIT (Massachusets Institute of Tecnology), Laboratorios Bell (ATT), General Electric, (en otras) se vieron en la necesitad de crear sistemas operativos que fueran multiusuario y multitarea. Inventaron el Unix.
Los inventores de Laboratorios Bell fueron Ken Thomsom, Dennis Richie (este es el inventor de C). Luego se hizo una version en la Universidad de Berkely. Sistemas que usan Unix:
    * Microsoft (Xenix)
    * Hewlett Packard (HP)
    * Apple (MacOS)
    * Sun Microsystems (Solaris)

Dentro de estos sistemas existen distintos tipos de unix.
De todos ellos nosotros nos concentramos en Linux. Linus Torvads.
Hay muchos sistemas Linux:
    * Ubuntu (el que tenemos aca)
    * Fedora
    * CentOS.
    * Red Hat dio origen a Fedora.
    * OpenSUSE.

### Componentes de Linux:
* Kernel: el manejo de memoria, de almacenamiento, y perifericos, etc.
* Shell: Este la interface entre el usuario y el sistema operativo. El prompt de donde se ejecutan los comandos.
