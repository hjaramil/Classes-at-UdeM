# Comparacion de archivos.
Vamos a ver los comandos

* ```diff```. Se tiene dos archivos ```A```, ```B```.
$B=A+ \delta$, entonces $B-A=\delta$. En este caso
$\delta$ es la salida del comando ```diff```.
Para recuperar $B$ de $A$, tenemos que $B=A - \delta$.

Como trabajan los programs de "version control" (ej GIT, SVC, Dropbox, vi )

$$A_1 \rightarrow A_2 \rightarrow \cdots \rightarrow A_{n-1} \rightarrow A_n .$$
Los deltas son $\delta_i = A_i - A_{i-1}$.


$$  A_n = A_{n-1} + \delta_n = A_{n-2} + \delta_{n} + \delta_{n-1} + \cdots = A_j + \sum_{i=j+1}^n \delta_i .$$

Si queremos recuperar $A_j$ del $A_n$, despejando tenemos

$$ A_j = A_n - \sum_{i=j+1}^n \delta_i .$$

El comando ```dif``` esta disnado para archivos de texto.

Algunas opciones importantes del ```diff```.

**Opciones**:
1. ```-i-```: Ignora si es mayuscula o minuscula
2. ```-w``` : Ignora espacios en blanco.
3. ```-r``` : Comparacion recursiva.
4. ```-B``` : Ignora lineas en blanco
5. ```-color```: color coded


* cmp: Compara dos archivos y si son binarios arroja un poco mas de informacion.

* kompare: GUI para comparar basado en el escritorio KDE.
