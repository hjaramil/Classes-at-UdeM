## Smooth curves (curvas suaves).
Notas en clase en el archivo ``.tex``.
### Actividad #1
Haga una curva con una flecha al final. El color de la curva es negro y el de la flecha azul.

### Actividad # 2
Sombree un circulo de arriba a abajo, de abajo a arriba, de adentro a fuera y de afuera adentro de dos formas.
* Intercambiando colores
* Intercambiando lugares.
