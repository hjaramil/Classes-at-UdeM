### Gradiente
El gradiente es el vector de derivadas parciales.
Si $f: \mathbb{R}^m \to R$,  el gradiente de $f$ es un **campo vectorial**

con $x=(x_1, x_2, \cdots, x_m)$
\begin{eqnarray}
\nabla f : \mathbb{R}^m &\to& \mathbb{R}^m \\
x &\mapsto& \nabla f(x) = \begin{pmatrix} \frac{\partial f}{\partial x_1} \\ \frac{\partial f}{\partial x_2} \\ \vdots \\ \frac{\partial f}{\partial x_m} \end{pmatrix}
\end{eqnarray}

#### Metodo 1
para calcularlo con `SymPy`

```python
from sympy.vector import CoordSys3D, gradient
C = CoordSys3D('')

x,y,z = symbols(' x y z')
f = C.x*sp.sin(C.y)*C.z**2
f
```

```python
gradf = gradient(f)
gradf
```

```python
gradfM = gradient(f).to_matrix(C)
gradfM
```

```python
# evaluemos
from sympy import pi, Matrix
gradfMs = gradfM.subs( [ (C.x,1), (C.y,pi/2), (C.z, 2)])
gradfMs
```

## Hessian
Dada una funcion $f=f(x_1, x_2, \cdots, x_n)$ la matriz Hessina se define como


$$ H_{ij} = \frac{\partial f}{\partial x_i \partial x_j} .$$

La matriz Hessiana se usa para verificar valores extremos en superficies o aproximaciones en optimizacion basadas en la expansion de Taylor.


```python
from sympy import hessian
x,y,z = symbols('x y z')
f = x*sp.sin(y)*z**2
v = list( ordered(f.free_symbols))
display(v)
hessian(f,v)
```

### Regla de la cadena
Asuma $f(x,y,z)$  con $x=x(t), y=y(t), z=z(t)$
podemos encontrar $df/dt$ usando la derivada total,

```python
from sympy import Function
x,y,z = symbols('x y  z', cls=Function)
f = symbols('f', cls=Function)
t = symbols('t')
x = x(t)
y = y(t)
z = z(t)
f = f(x,y,z)
f
```

```python
diff(f,t)
```

```python
# agregamos un parametro mas, las derivadas se convierten en partial
x,z,y = symbols('x y z ', cls=Function)
f = symbols('f', cls=Function)
s = symbols('s')
x = x(s,t)
y = y(s,t)
z = z(s,t)
f=f(x,y,z)
f
```

```python
# regla de la cadena para derivadas parciales
diffs = diff(f,s)
diffs
```

```python
# evaluemos
fst = diffs.subs(f, x**2*y)
fst
```

```python
fst2 = fst.doit()
fst2
```

```python
# mas sustituciones
fst2.subs( {x:s**2+t, y:s**2-t**2})
```

```python
dfds =fst2.subs( {x:s**2+t, y:s**2-t**2}).doit()
dfds
```

```python
# sustitucion final numerica
dfds.subs( {s:5, t:2})
```

## Valores extremos
Asuma $f=f(x,y)$. Los extremos de la funcion ocurren en:

* Puntos en la frontera
* Puntos donde $\nabla f = 0.$

Debemos verificar si estos puntos son de **maxima**, **minima**, o **silla**.

La verificacion se hace mediante la matriz Hessiana


\begin{eqnarray}
H = \begin{pmatrix} f_{xx} & f_{xy} \\ f_{yx} & f_{yy} \end{pmatrix}
\end{eqnarray}
donde $f_{xy}=\frac{\partial^2 f}{\partial x \partial y}$ y asi con los demas.  En general esta matriz es simetrica.
Llamamos $D$ al **discriminante** de la matriz Hessiana, es decir al determinante.


$$D = \det H = f_{xx} f_{yy} - f_{xy}^2  .$$
Un punto $(a,b)$ se dice que es

* **maximo local** si $f_{xx}<0$ y $D>0$.
* **minimo local** si $f_{xx}>0$ y $D>0$.
* **punto de silla** si $D<0$.
* **sin conclusion** si $D=0$.
