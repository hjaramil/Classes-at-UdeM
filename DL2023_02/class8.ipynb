## Convolucion (Continuacion)
El continuo existe el Delta de Dirac $\delta(t)$ que vale $\infty$ $t=0$ y $0$ para $t \ne$. pero $\int_0^{\infty} \delta(t) dt = 1$. Esto pertenece a la teoria de distribuciones (dentro del analisis funcional).

La respuesta impulso a un sistema lineal la notamos como
$$ g(t ) = L (\delta(t))$$
Asuimos que el el sistema $L$ es lineal, $L(a f(t) + b g(t))= a L(f(t)) + b L(g(t))$ e invariante con el tiempo, es decir que si $g(t) = L( \delta(t))$ entonces $g(t - \tau)= L(\delta(t - \tau)]$

Entonces suamos la propiedad de localicacion del delta de Dirac

$$ f(\tau) = \int_{-\infty}^{\infty} f(t) \delta(t - \tau) d t .$$

Veamos entonces la convolucion $z(\tau)$.

$$z(\tau) = L( f(\tau)) = L \left [   
\int_{-\infty}^{\infty} f(t) \delta(t - \tau) d t
    \right ] = \int_{-\infty}^{\infty} f(t) L(\delta(t-\tau) dt = \int_{-\infty}^{\infty} f(t) g(t - \tau) dt $$


De aca sale el concepto de **convolucion** y escribimos

$$ (f \ast g)(\tau) = \int_{-\infty}^{\infty} f(t) g(t - \tau) d t  \tag{1} .$$

Por la pregunta de Leandro:

* La **convolucion** resulta de forma natural de aplicar filtros lineales e invariantes en tiempo, tales como  en la tierra cuando se propaga una onda sismica. Ejemplo a mano alzada.

* Si le pregunta a un matematico, le va a decir que la convolucion es una operacion que te permite varias cosas

** Resolver ecuaciones diferenciales.
** Usando la transformada de Fourier (de Laplace). Por esta convierte una ecuacion diferencial en una ecuacion algebraica y por que la convolucion ante una transformada de Fourier se convierte en un producto.

$$ \mathcal{F}(g'(t)) = \text{i} \omega F(\omega) .$$
De esta forma una ecuacion diferencial (izquieda) se convierte en algebraica (a derecha)

$$ \mathcal{F}(f \ast g) = F(\omega) G(\omega).$$

* Si le pregunta a un estadistico va a decir que la probabilidad conjunta de la suma de dos variables aleatorias $X, Y$ es la convolucion de las mismas $$P(X + Y) = P(X) \ast P(Y) .$$

# Metodos de descenso con gradiente
## Metodos generales de descenso a un minimu.
No todos los metodos usan el gradiente para descender.

**Comentarios**:
* **Simulating Annealing**: Es buena tecnica para en contrar el minimo (maximo) global. Global es en todo el tamano de los datos. Pero es lento
comparado a los metodos de gradiente. Simulating Annealing no usas derivadas. (gradiente=vector de derivadas parciales)  

* **Pattern Search** (busqueda de patrones). Tampoco usa derivadas pero se usa en ANN: Artificial Neural Networks.

* **Metodos de Newton**: Usan derivadas y normalmente necesitan el Hessian (la matriz de segundas derivadas $\partial^2 f/ \partial x_i \partial x_j$). Este es muy costoso, sobre todo para problemas grandes como los de ANN, DL. No los usamos

* **Quasi-Newton**: Estos son una forma rapida e inexacta de los metodos de Newton donde no se tiene que calcular el Hessian. El mas importante es el L-BFGS: Limited Memory de autores Broyden-Fletchere-Goldfrab-Shanno.  
Mas que todo se centra en la diagonal de Hessiano. No los describimos aca pero son muy usados en ANN y en general. Muy populares.

* **Algoritmos Evolutivos**: Son inspirados por la evolucion biologica. No requieren calculos de derivadas. Hay dos muy famosos en esta categoria: Particle Swarm Optimization, Genetic Algoritms (algoritmos geneticos)
Son buenos para encontrar maximo (minimo) absoluto pero son lentos en su convergencia, entonces no se usan mucho en DL.

* **Metodos de descenso con gradiente**: (este el que usamos en el curso)
** **Steepest Descent**: Es lento, intuitivo, es analitico para funciones cuadraticos. Lo veremos en la proxima clase.
** **Conjugate Gradient (CG)**:  Gradiente conjugado. Es el mejor para resolver el problema $Ax = b$. Recuerde que resolver un problema lineal es lo mismo que minimizar un cuadratico. Ver notas a mano alzada. En matrices enormes y no densas (sparse matrices). Pero no sirve para problemas no lineales.
** **Non-linear conjugate gradient**. Una extension del anterior para funciones objetivo no cuadraticas. El L-BFGS cae en esta cateroria.
** **Stochastic Gradient Descent**: Metodo de descenso con gradiente estocastico. Es el mismo metodo de gradiente pero se usa funciones aleatorias para mejorar la velocidad de proceso. En la proxima clase veremos ejemplos donde el calculo del gradiente y el uso es enorme en ANN y no es posible en estos momentos, por problemas de memoria, usarlo.
