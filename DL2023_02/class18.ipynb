# Desarrollo de un modelo (Development of a model)
Para desarrollar un modelo en ML y particularmente en DL necestiamos cubrir los siguientes pasos:

* Obtener los datos.
* pre-procesar los datos.
* construir la arquitectura.
* entrenar y probar el modelo.
* hacer las predicciones.

## Preproceso de los datos.
Una vez obtenemos los datos los debemos procesar por varias razones. Por ejemplo,

* Homogenizacion: Se tiene una cantidad de imagenes. Es bueno que todas ellas tengan el mismo tamanho (en pixeles, por decir algo 32 x 32, en el caso de CIFAR).
* Formateo: Los datos deben estar en un formato amigable para el software que se usa.
* Normalizacion: Recuerde la clase de ML, que estudiamos distintos tipos de normalizacion.
* Limpiar numeros como los NaN (Not a number), NONE, etc.

## Entrenamiento del modelo.
Es importante entender que los datos se deben dividir en varios grupos, unos de entrenamiento y otros de prueba. En la clase de ML, hablamos del k-fold, pero se pueden dividir los datos en solo dos grupos 70% para entrenamiento y 30% para prueba. Algunas veces
se usan 3 grupos de datos: training, crossvalidation, testing que pueden ser del orden de 40%, 30% , 30%.



## Prediccion.
Una vez el modelo esta "bien" entrenado, se usa para hacer predicciones. Este es el objetivo final.

Todos los pasos que se muestran arriba, no necesariamente se corren un vez. Puede haber un ciclo, en donde si la prueba (testing) no es exitosa, toca regresar a: tomar nuevos datos,
cambiar la arquitectura (mas nodos, o menos nodos, o mas capas, etc).

# Deep Learning (DL): Aprendizaje Profundo
Veamos unos diagramas para comprender el leguaje que queremos usar cuando hablamos de DL.

image.png

La figura muestra una red neuronal con 3 capas, entrada, oculta y salida.  Esta red es "fully connected" or "dense" (conectada en su totalidad, o densa). Generalmente los nodos solo albergan escalares.

Por que agregue esta grafica?  En MLP: Multilayer Perceptron, unas pocas capas son suficientes. Para DL pueden haber 100 capas o mas y estos dibujos son insuficientes.

Necesitamos otra forma de representar las arquitecturas de las redes neuronales profundas.

image.png

El dibjuo de arriba es mas apropiado para redes profundas, donde
hay muchas capas o muchos nodos, o ambos.  Este es tipico en CNN: Convolutional Neural Networks (este tema lo comenzamos en la proxima clase).
Una explicacion a la grafica, rapida.

* Al comeinzo ( la caja grande y delagada, como una caja de Pizza)
representa los datos de entrada. Los numeritos indican las dimensiones.

* Luego una caja mas pequenha. Este el filtro (recuerde la convoluicion).

* La caja marron pegada del filtro se llama "pooling layer".
De nuevo, esto lo explicamos la semana entrante.

* Sigue una sucesion de capas con o si pooling layer.

* Despues, casi al final, dos "varillas largas y delgadas". Estas son "fully connected"

* La barra morada al final ya es salida con una clasificacion, en este caso "soft".

Esta arquitectura, particularmente, se llama el AlexNet.

El software que genero la figura de arriba es:

[PlotNeuralNet](https://github.com/HarisIqbal88/PlotNeuralNet).
