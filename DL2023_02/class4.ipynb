Si un tensor de rango 3 es totalmente antisimetrico entonces
es de la forma

$$ a_{ijk} = \epsilon_{ijk} a_{123} \tag{1}$$
Veamos por que.  

* Si 2  o 3 de los indices son repetidos entonces $a_{ijk}=0$ (por que es antisimetrico). Por ejemplo $a_{112}$, entonces $\epsilon_{112}=0$, entonces $a_{112} = \epsilon_{112} a_{123}=0$. Cumple la igualdad.
Solo quedan dos casos

* $(i,j,k)$ sea $(1,2,3) \; o \;  (2,3,1) \; o  \; (3,1,2)$.  En este caso tenemos que como $a_{ijk}$ es totalmente antisimetrico
$a_{123}=a_{231}=a_{312}$ (se puede probar con la definicion. Es decir cambiando los indices $ij$ por $ji$ (a esto se le llama un transposicion) y se deja como ejercicio esta verificacion. Ademas $\epsilon_{123}=\epsilon_{231}=\epsilon_{312}$, de forma que $a_{ijk}= \epsilon_{ijk} a_{123}$.

* El ulimo caso es $(i,j,k)$ se impar, es decir $(1, 3, 2)$ o $(3,2,1), o $(2,3,1). En este caso $a_{ijk} = - a_{123}$ por que $(1,2,3)$ es par y las demas en este item son impares y $\epsilon_{ijk}=-1$, Entonces de nuevo
$a_{ijk}= \epsilon_{ijk} a_{123}$.

Queremos generalizar el simbolo de Levi-Civita a dimensiones superiores, es decir $n > 3$. Para esto necesitamos definir unos cuntos conceptos.


Definicion (**permutacion**): Recuerde la definicion de $I_n=\{ 1, 2, \cdots, n \}$i. Una permutacion $P_n$ es una biyeccion en $I_n$.

Es decir, por ejemplo, en $I_3$. Tenemos las siguientes permutaciones:



\begin{eqnarray*}
  1 \mapsto 1   \quad &,& \quad 1 \mapsto 2  \quad , \quad 1 \mapsto 3 \quad , \quad
  1 \mapsto 1   \quad , \quad 1 \mapsto 2  \quad , \quad 1 \mapsto 3 \\
  2 \mapsto 2   \quad &,& \quad 2 \mapsto 3  \quad , \quad 2 \mapsto 1 \quad  , \quad
  2 \mapsto 3   \quad , \quad 2 \mapsto 1  \quad , \quad 2 \mapsto 2  \\
  3 \mapsto 3   \quad &,& \quad 3 \mapsto 1  \quad , \quad 3 \mapsto 2  \quad  , \quad
  3 \mapsto 2   \quad , \quad 3 \mapsto 3  \quad , \quad 3 \mapsto 1  .
\end{eqnarray*}

A las permutaciones se les llama comunmente como $\sigma$, o $\pi$, pero aca usamos $P_n$.  Uno puede permutar objetos. Sea $a_{ijk}$ un objeto, se permutan los indices con estas formulas y esto permuta los objetos.

Defincion (**transposicion**): Data una permutacion podemos reversar (intercambiar) el orden en solo 2 de las componentes. A esto se le llama **transpocicion**.
Por ejemplo Si en la permutacion $(3 \; 2 \; 1)$  intercambiamos la primera con la tercera componente obtenemos $(1 \; 2 \; 3)$.
Nosotros podemos "navegar" desde la identidad (la identidad es aqueya permutacion que no hace nada. Es decir $(1 \; 2 \; \cdots \; n)$) hasta cualquier permutacion con solo transposiciones.


Ejemplo: para movernos de $(1 \; 5 \; 3 \;2 \; 4)$ desde la identidad hacemos las siguientes transposiciones.


$$ ( 1 \; 2 \; 3 \; 4 \; 5 ) \mapsto (1 \; 5 \; 3 \; 4 \; 2) \mapsto (1 \; 5 \; 3 \; 2 \; 4).  $$

Si para ir desde la identidad a una permutacion $P_n$ tenemos hacer un numero par de transposiciones, decimos que la permutacion es par **par**. Si al contrario necesitamos un numero impar de permutaciones decirmos que la permutacion $P_n$ es **impar**. Las permutaciones que hicimos arriba $T_{25}$ y $T_{45}$, ahora podemos hacer   $I \mapsto T_{25} \mapsto T_{25}=I$ por que se hizo y deshizo. Existe un camino mas corto para ir de la identidad a $P_n$ y cualquier otro camino mas largo se le suma $2 k$ donde
$k$ es el numero de veces que "hizo y deshizo". Hay consitencia: par+par da par, impar + par = impar.

Generalicemos el Levi-Civita para $n$ dimensiones:

\begin{eqnarray}
\epsilon_{i_1 i_2 \cdots i_n} = \left \{
    \begin{array}{cc}
    1 & \text{si   } P=(i_1 i_2 \cdots i_n) \quad \text{es par} \\
    -1 & \text{si   } P=(i_1 i_2 \cdots i_n) \quad \text{es impar} \\
    0 & \text{si  hay por lo menos dos indices repetidos }
    \end{array}
    \right .
\end{eqnarray}

Definicion (**determinante**): Asuma $i_1, i_2, i_3, i, j \in I_3$. Entonces

$$\det(a_i^j) = | a_i^j | = \epsilon_{i_1 i_2 i_3} a_1^{i_1} a_2^{i_2} a_3^{i_3}.  $$
La generizacion del determinante a $n$ dimensiones (es decir, $I_n$), es la siguiente

$$\det(a_i^j) = | a_i^j | = \epsilon_{i_1 i_2 \cdots i_n} a_1^{i_1} a_2^{i_2} \cdots  a_n^{i_n}.  $$


Como una curiosidad, esta suma multiple (hay $m$ sumas o signos $\Sigma$)
usando la teoria combinatoria tiene $n^n$ terminos, de ellos solo unos "pocos" ($n!)$ no son cero (las permutaciones de $I_n$). De forma que
$n^n - n!$ son ceros.

Otra aplicacion del Levi-Civita es en el producto cruz $\times$.

Definicion (**producto cruz**): Asuma $i_1, i_2, i, j, k \in I_3$,
Entonces si tenemos dos vectors $A = (a^i)$, $B=(b^i)$, el producto cruz de estos dos vectores se define como

$$ (A \times B)_k  =  \epsilon_{k i_1 i_2} a^{i_1} b^{i_2} .$$
Queda como tarea verificar que esto es asi.






Que podria haber motivado una definicion asi. Vamos compara el producto cruz con el determinate.

Por ejemplo:

\begin{eqnarray}

a^1 & a^2 & a^3 \\
b^1 & b^2 & b^3 \\
\end{array}
\right | = \epsilon_{ijk} \hat{e}_i a^j b^k.
\end{eqnarray}
donde $\hat{e_1} = (1, 0, 0)^T$, ,$\hat{e_2}=(0,1,0)^T$,
$\hat{e_3}=(0,0,1)^T$.

De forma que podemos escribir para $A=(a^1, a^2, a^3)$ y $B=(b^1, b^2, b^3)$,

$$ (A \times B)_i = \epsilon_{ijk} a^j b^k . $$


Quisieramos examinar el siguiente objeto


$$ \epsilon_{ijk} a_r^i a_s^j a_t^k $$
para $i,j,k,r,s,t \in I_3$ que es un tensor de rango 3.

Vamos a mostrar que este tensor es completamente anti-simetrico.
Es decir que si reversamos dos de los indices se cambia el signo solamente.
Los indices (libres) son $r,s,t$. Tendriamos que usar 3 veces el algortimo: $r,s$ por $s,r$, $r,t$ por $t,r$ y $s,t$ por $t,s$. Solo hacemos el primero y dejamos los otros dos al estudiante. Es decir,
vamos a probar que si cambiamos $r,s$ por $s,r$ se cambia el signo y nada mas pasa.  Un tensor es totalmente antisimetrico si al hacer cualquier transposicion se cambia el signo.
Vamos

$$\epsilon_{ijk} a_r^i a_s^j a_t^k = \epsilon_{ijk} a_s^j a_r^i a_t^k =-\epsilon_{jik} a_s^j a_r^i a_t^k=-\epsilon_{ijk} a_s^i a_r^j a_t^k$$

Donde en la primera igualdad utilizamos la ley conmutativa del producto. Una de las fortalezas de la notacion indicial es que todo conmuta.
En la segunda igualdad hicimos una transpocicion al Levi-Civita.
En la tercera igualdad, dado que $i,j,k$ son "dummies" puedo renombrarlos como quiera. Y aca a la $i$ la llamo $j$ y a la $j$ la llamo $i$.

Vemos que al reversar $r,s$ por $s,r$ se cambio el signo. Igual con los otros dos "reversasos".

Vemos entonces que el tensor
$ \epsilon_{ijk} a_r^i a_s^j a_t^k $ es totalmente antisimetrico en los indices $r,s,t$. Recuerde la ecuacion.
