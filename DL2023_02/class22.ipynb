# AlexNet
Esta red gano el concurso
[ILSVRC](https://www.image-net.org/challenges/LSVRC/#:~:text=The%20ImageNet%20Large%20Scale%20Visual,image%20classification%20at%20large%20scale.)

ImagNet Large Scale  Vision Recognition  Challenge. Gano este concurso en 2012.
Autores: Alex Krizhevsky, Ilya Sutskever, Geoffrey E. Hinton, de la Universidad de Toronto.
En el concurso se usaron 1.2 millones de imagagenes para el entrenamiento, 50 mil para la validacion y 150 mil para las pruebas, las clases fueron 1000.

Fue la primera red que incluyo GPU. Usaron dos redes una pegada con otra para tener dos sistemas GPU (a la par).

La arquitectura de la AlexNet es:
Lista corta y luego el detalle.
Comparacion con la LeNet5

* Redes convolucionales: LetNet5 tiene 3 y Alexnet tiene 5
* Funciones de Activacion: LeNet5 usa tanh (nosotros implementamos una ReLU), AlexNet usa ReLU. Al final ambas usan la softmax

* Pooling: LetNet5 uso AveragePooling, AlexNet usa MaxPooling.

*  AlexNet usa dropout (son los mismo autores).

Detalle de la red:

* Capa de entrada y convolucion 1:  La entrada tiene 224 x 224 x 3 pixeles (RGB). La convolucion tiene los siguientes parametros.
    * no padding
    *  stride s=4
    * kernel 11 x 11 con 96 filtros de salida
Contemos: (224-11)/4+1 = 52.24  (se trunca). La salida tiene
54 x 54 x 96. El numero de parametros

$$\text{learning parameters}=(11 \times 11 \times 3 + 1) \times 96 =34944 .$$

Se usa ```batch normalization``` en todas las capas. Esto no aparece en la publicacion original.
* MaxPooling: parametros
    * kernel $3 \times 3$
    * Stride $s=2$.
    * Padding=None
La salida tiene tamano
$$ (54 -3)/2 + 1 = 26.5.$$, truncando, la salida debe ser
de $26 \times 26 \times 96.$

* Convolucion 2: Parametros
    * kernel $5 \times 5$
    * Stride $s=1$
    * Filtros 256
    * Padding 1
Tamano de salida $(25-5 + 2(2))/1+1=26.5$ El tamano es
$$ 25 \times 26 \times 256.$$
Parametros
$$\text{learning parameter} = (5 \times 5 \times 96 +1) \times 256 = 613656 .$$

* MaxPooling:
    * Kernel $3 \times 3$.
    * Stride $s=2$.
    * padding:none
Tamano $(26-3)/2+1 = 12.5$ se trunca a 12 y la salida tiene tamano $12 \times 12 \times 256$.

* Convolucion 3:
    * kernel $3 \times 3$.
    * Stride $s=2$
    * padding none
    * 384 filtros de salida
Tamano de salida $(26-3)/2 + 1= 12.5$ se trunca a 12
$$12 \times 12 \times 256 .$$
Numero de parametros
$$\text{numero de parametros} = (3 \times 3 \times 256 +1) \times 384 = 885120.$$

* Convolucion 4.
    * kernel $3 \times 3$.
    * Stride $s=1$
    * padding $p=1$
    * filtros 384
Tamano de salida $(12-3+2)/1 +1 = 12 $
La salida es $12 \times 12 \times 384$.
$$\text{numero de parametros}= (3 \times 3 \times 384 +1) \times 384 = 1327488 .$$

* Convolucion 5.
    * kernel $3 \times 3$.
    * Stride $s=1$
    * padding $p=1$
    * filtros 256
Tamano de salida $(12-3+2)/1 +1 = 12 $
La salida es $12 \times 12 \times 256$.

$$\text{numero de parametros}=(3 \times 3 \times 384 + 1) \times 256 = 884992 .$$

* MaxPooling:
    * Kernel $3 \times 3$.
    * Stride $s=2$.
    * padding:none
Tamano de salida $(12-3)/2+1= 5.5$ truncamos a 5 y la salida tiene tamano
$ 12 \times 12 \times 256 $

* Dropout regularization rate 0.5. Activacion ReLU en todas las capas.

* Flatten. Output size $5 \times 5 \times 256=6400$.

* Fully connected. Input 6400 , output 4095
$$\text{numero de parametros}=(6400+1) \times 4096 = 2628496 .$$

* Fully Connected. Input 4096 el output 6400
$$\text{numero de parametros}=(6400+1) \times 4096 = 2628496 .$$

* Fully Connected. Input 4096 el output 1000 (para nosotros el ouptut=17).

$$\text{numero de parametros}=(4096+1) \times 17=69649 .$$
