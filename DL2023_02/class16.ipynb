# Multilayer Perceptron (MLP)
La idea, originalmente de Rosenblat, es usar mas de una capa (aparte de la entrada y la salida). Entre la entrada y la salida debe haber una capa oculta.


image.png

Matematicamente esta red neuronal, que se llama MLP, tiene como capa de entrada los datos $x=(x_0, x_1, \cdots, x_4)$.
La capa del medio se llama de activacion, la vamos a construir.
Esta capa tiene los valores $a = a_0^{(2)}, a_1^{(2)}, \cdots, a_5^{(2)}$. La notacion en este caso es subindices para las componentes del vector (tensor) y el superindice $^{(2)}$ indica el numero de la capa donde la capa de entrada es la capa $^{(1)}$.
Tenemos un solo nodo de salida.  A los nodos tambien se le concocen como **neuronas**.

Lo iportante es entender como curcula la informacion en una red neuronal. Hay dos tipos de movimiento.


* Forward: Hacia adelante, que explicamos en un momento
* Backward: Hacia atras (desde la ultima hasta la primera capa). Que se llama **backpropagation**.

Para comenzar, antes de definir las activacion en la capa $^{(2)}$, cremos un paso intermedio donde se hace un **promedio ponderado** de los datos de entrada con la matriz de pesos $\theta$.  Matematicamente, este promedio ponderado lo podemos escribir como


$$ z_i^{(2)} = \sum_{j=0}^4 x_j \theta_{ji}^{(1)} \quad , \quad i=1,2,3,4,5 .$$
La pregunta es: quien es $a$?. La activacion. La activacion es una funcion que toca definirla como un **hiperparametro**.

For ejemplo, una funcion que fue muy popular en los 90s es la **logistic**

$$ a_i^{(2)} = f(z_i^{(2)}) = \frac{1}{1 + \exp(-z_i^{(2)})} \quad , \quad i=1,2,3,4,5 .$$


Como llegamos a la neurona roja (final). Solo tenemos una y
generamos primero el promedio ponderado, y luego la activacion.
Ese promedio ponderado, lo llamamos $z^{(3)}$,

$$z^{(3)} = \sum_{j=0}^5 a_j^{(2)} \theta_j^{(2)}  .$$
Aplicamos la activacion.

$$ h(\Theta^{(1)},\Theta^{(2)}, x) = a^{(2)} = f(z^{(3)}) = \frac{1}{1 + \exp(-z^{(3)})} .$$

Podria pasar que tivieramos una clasificacion multiclase. Por ejemplo en imagenes determinar si el resultado es un cerdo, un gato, un perro, etc.

## Arquitectura de Redes.
La arquitectura muestra como se estructura una red neuronal artificial. Para esto se necesitan parametros e hiperparametros.
Los parametros son lo pesos $\theta$, los demas se llaman hiperparametros. Tenemos por ejemplo,

* Numero de capas.
* Numero de nodos por capa.
* Activaciones (funciones de activacion, proxima seccion)
* Funcion de costo. Para optimizar los pesos. Esta al final de la red neuronal y se usa para la "backpropagation"
* Tecnicas de optimizacion. Gradiente con aceleradores (momentum, Nesterov, Adam, AdaGrade, etc)
* Metricas de precision. (se ve en ML, accuracy, precision, recall, F1, etc.)

En adicion a esto, hay hiperparametros importantes, por ejemplo, la forma como se conectan los nodos de las redes.
Por ejemplo.

* Fully connected: densamente conectados. El caso de las MLP.
* CNN: Convolutional (correlational) neural networks. Estas solo concectan unas pocas neuronas (el tamano del filtro) entre si. Esto reduce drasticamente el tamano de las redes.  
* RNN: Recurrent Neural Networks. Aca la salida le sirve de "feedback" (retroalimentacion) a la entrada. Las vemos mas adelante.
* GANs: Generative Adversarial Networks: Aca hay un sistema de dos redes que interactuan. El falsificador vs el detective.




# Funciones de Activacion.
Recuerde que inicialmente se hace un promedio ponderado de la entrada con los pesos (multiplicacion de matrices, o tensores producto interno).

Que pasa si no usamos funciones de activacion:
En nuestro ejemplo
Inicialmente calculamos
$$ x^T \Theta^{(1)}.$$
Como esto da un vector, fila, mejor lo volteamos para volverelo vector columna
$$ (\Theta^{(1)})^T x .$$

Como paso siguiente (no hay activacion), repetios la ponderacion con este vector de salida, ahora como entrada con los siguientes pesos.

$$(\Theta^{(2)})^T ( \Theta^{(1)})^T x ,$$
y asi suscesivamente.  En general para $L$ capas

$$ h(\Theta, x) = \prod_{\ell = 0}^{L-1} (\Theta^{L - \ell})^T x
= (\Theta^{(L-1)}) \cdots (\Theta^{(2)}) (\Theta^{(1)}) x  \tag{1}.$$

Si lo hiciermos asi, este es un problema lineal, por que al multiplicar todas las matrices, da otra matriz, y el problema al final seria de la forma $h=A x$, que es un problema lineal.

Con enfoque lineal solo se pueden resolver problemas lineales. Entonces como resolvemos problemas mas complicados que estan lejos de ser problemas lineales?

Si las activaciones fueran la identidad $a(x)=x$, no pasa nada, y tenemos la red descrita, matematicamente, en la ecuacion (1).

Veamos ejemplos de funciones de activacion.

## Heaviside (Step, funcion de paso)

\begin{eqnarray}
f(x) = \left \{
    \begin{array}{cc}
    1 & \text{si    } x \ge 0 \\
    0 & \text{si    } x < 0  \\
    \end{array}.
    \right .
\end{eqnarray}


````python
import numpy as np
import matplotlib.pyplot as plt

def H(x):
    if x>=0 :
        return 1
    else:
        return 0

xdomain = np.arange(-5,5, 0.01)
y = [H(x) for x in xdomain]


plt.xlabel("x")
plt.ylabel("y")
plt.title("Heaviside step function")
plt.plot(xdomain, y)
````

## Logistic Activation function.


$$f(x)= \frac{1}{1 + \exp(-x)} .$$

````python
def s(x):  # sigmoid
    return 1/(1 + np.exp(-x))

xdomain = np.arange(-5,5, 0.01)
y = [s(x) for x in xdomain]


plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmoid step function")
plt.plot(xdomain, s(xdomain), label="simple")
plt.plot(xdomain, s(4*xdomain), label="4x")
plt.plot(xdomain, s(12*xdomain), label="12x")
plt.legend()
````

Se sugiere que $$H(x) = \lim_{n \infty} s(nx).$$

## Tangente hiperbolica
$$ f(x) = \tanh (x).$$

````python
def t(x):  # sigmoid
    return np.tanh(x)

xdomain = np.arange(-5,5, 0.01)


plt.xlabel("x")
plt.ylabel("y")
plt.title("Tangente hiperbolica step function")
plt.plot(xdomain, t(xdomain), label="simple")
plt.plot(xdomain, t(4*xdomain), label="4x")
plt.plot(xdomain, t(12*xdomain), label="12x")
plt.legend()
````

La tangente hiperobolica se parece a la sigmoid, pero va entre -1, y 1. Se sugierer $\lim_{n \to \infty} \tanh(nx) = signo(x)$

## Activacion ReLU
\begin{eqnarray}
\text{ReLU(x)} = \left \{
    \begin{array}{cc}
    x & \text{si    } x \ge 0 \\
    0 & \text{si    } x < 0  \\
    \end{array}.
    \right .
\end{eqnarray}


## Leaky Rectified Linear Unit
\begin{eqnarray}
\text{leaky ReLU} (x) = \left \{
    \begin{array}{cc}
    x & \text{si    } x \ge 0 \\
    a x & \text{si    } x < 0  \\
    \end{array}.
    \right .
\end{eqnarray}
donde $a$ es pequenho, por ejemplo $a=0.001$


## ELU
\begin{eqnarray}
\text{ELU}(x) = \left \{
    \begin{array}{cc}
    a \exp(x - 1)  & \text{si    } x \ge 0 \\
    x & \text{si    } x < 0  \\
    \end{array}.
    \right .
\end{eqnarray}
Si $a=1$ la funcion y su derivada son continuas.


````python
def ReLU(x):
    if x >= 0 :
        return x
    else:
        return 0

xdomain = np.arange(-5,5,0.01)
y=[ReLU(x) for x in xdomain]

plt.xlabel('x')
plt.ylabel('y')
plt.title("ReLU activation Function")
plt.plot(xdomain, y);
````

````python
a=0.01
def LReLU(x):
    if x >= 0 :
        return x
    else:
        return a*x

xdomain = np.arange(-5,5,0.01)
y=[LReLU(x) for x in xdomain]

plt.xlabel('x')
plt.ylabel('y')
plt.title("Leaky ReLU activation Function")
plt.plot(xdomain, y);
````

````python
a=1
def ELU(x):
    if x >= 0 :
        return a*(np.exp(x) - 1)
    else:
        return x

xdomain = np.arange(-5,5,0.01)
y=[ELU(x) for x in xdomain]

plt.xlabel('x')
plt.ylabel('y')
plt.title("ELU activation Function")
plt.plot(xdomain, y);
````
