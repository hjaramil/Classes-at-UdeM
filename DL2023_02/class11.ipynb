# Gradiente con descenso estocastico: Continuacion
Tres preguntas:

1. Que pasa al interior de la region de confusion
2. Podemos estimar la incertidumbre dentro de esta region
3. Como podemos estimar el parametro $\eta$ de la ecuacion de descenso.

Vamos a formular la estadistica del problema.
$$ \mathbb{E} [ \nabla C_i(\theta)] = \frac1m \sum_{i=1}^m \nabla C_i(\theta) = \nabla C(\theta) .$$

Que tal si la distribuciom estadistica de los gradientes no es uniforme. Asume una densidad de probabilidad $\mu_i$ para la muestra $i$.


$$ \mathbb{E} [ \nabla C_i(\theta)] = \sum_{i=1}^m \mu_i \nabla C_i(\theta) = \nabla C(\theta) .$$
Si conocemos la distribucion estadistica de los gradientes podemos usar $\mu_i$ en vez de $1/m$.

Otra unidad importante para estimar las estadisticas de mis gradientes es la varianza. La varianza (si lo ve) es como un como un mean square error (MSE).  



$$ \mathbb{V}= \frac{1}{m-1} \sum_{i=1}^m
 \| \nabla C_i(\theta) - \nabla C(\theta) \|^2.$$
Si la varianza es pequena el error de aproximar $\nabla C_i$ como si fuerea $\nabla C(\theta)$ es pequenho y el metodo de SGD se justifica. Si todos los gradientes locales son iguales el error es 0 (la varianza es 0) pero solo ocurre si la la funcion de costo es lineal y las funciones de costo **nunca** son lineales.
Esta varianza resuleve la pregunta (2) acerca de la incertidumbre.

Antes de visitar la pregunta (3) vamos a comentar algo.

En este momento hay dos extremos:
1. Descenso en gradiente pleno. Donde se calcula el gradiente para **todas** las muestras (SD). A este se le llamam el **batch**
2. Descenso con gradiente estocastico, donde solo se calcula el gradiente para **una** muestra.

Surge otra pregunta. Podemos pararnos en alguna parte en el medio?
La respuesta es si y el metodo se conoce como

**Mini batch**: En vez de calcular el gradiente en una sola muestra o en todas, se calcula en un "grupito". Esto reduce la varianza de arriba.  Hay ganancia, si en vez de calcualr el gradiente para $m$ muestras se calcula para $k$ muestras ($k \ll m$). La ganancia (velocidad de computo) es $m/k$.
