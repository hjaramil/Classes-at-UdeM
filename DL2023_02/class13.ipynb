# Continuacion en descenso con gradiente y momentum.
## Gradient Descent
The traditional recursion equation is $\theta^{(i)} = \theta^{(i-1)} - \eta \nabla f(\theta^{(i-1)})$.

```python
import numpy as np
import matplotlib.pyplot as plt
# easy function
def f(x):
    return x**4

# gradient
def gradf(x):
    return 4*x**3


# hard coded eta
eta=1.9999 # grandisimo, para ver los rebotes de pared
eps=10**(-6)

# initial point
x = 0.5

# list of points
xlist = [x]
ylist = [f(x)]
gradList = []

M=10 #epochs
for i in range(M):
    x = x - eta*gradf(x)
    # if abs(f(x)) < eps:
     #   break
    xlist.append(x)
    ylist.append(f(x))
    gradList.append(gradf(x))




# cast lists into numpy array
xarr = np.array(xlist)
yarr = np.array(ylist)

xdomain = np.arange(-0.5, 0.51,0.01)
plt.title("Steepest Descent")
plt.plot(xdomain, f(xdomain));
plt.quiver(xarr[:-1], yarr[:-1], xarr[1:]-xarr[:-1], yarr[1:]-yarr[:-1], \
           scale_units='xy', angles='xy', scale=1)

plt.scatter(xlist, ylist, s=80, facecolors='none', edgecolors='k');
plt.savefig("sd.pdf")
plt.show();
```

## Momentum (Version 1:Poliak)  $\theta^{(i)} = \theta^{(i-1)} - \eta \nabla f(x) + \beta ( \theta^{i} - \theta^{(i-1)} )$


```python
beta = 0.8
# initial point
x = 0.5

# list of points
xlist2 = []
ylist2 = []
gradList2=[]

for i in range(M):
    xlist2.append(x)
    ylist2.append(f(x))
    gradList2.append(gradf(x))
    x = x - eta*gradf(x) + beta*( xlist2[i] - xlist2[i-1])
    # if abs(f(x)) < eps:
     #   break


xdomain = np.arange(-0.5, 0.51,0.01)
# cast lists into numpy array
xarr = np.array(xlist)
yarr = np.array(ylist)

xdomain = np.arange(-0.5, 0.51,0.01)
plt.plot(xdomain, f(xdomain));
# plt.plot(xlist,ylist, color='b');
plt.quiver(xarr[:-1], yarr[:-1], xarr[1:]-xarr[:-1], yarr[1:]-yarr[:-1], \
           scale_units='xy', angles='xy', scale=1, label="steepest descent", color="blue")

# cast list into np arrays
xarr2 = np.array(xlist2)
yarr2 = np.array(ylist2)
xdomain = np.arange(-0.5, 0.51,0.01)
plt.plot(xdomain, f(xdomain));
plt.plot(xlist,ylist, color='blue');
# plt.quiver(xlist2,ylist2, 0.1, 0.1, color='brown');
plt.quiver(xarr2[:-1], yarr2[:-1], xarr2[1:]-xarr2[:-1], yarr2[1:]-yarr2[:-1], scale_units='xy', \
           angles='xy', scale=1, label="steepest descent + momentum ", color="brown")
plt.scatter(xlist, ylist, s=80, facecolors='none', edgecolors='k');
plt.scatter(xlist2, ylist2, s=80, facecolors='none', edgecolors='r');

plt.legend()

plt.savefig("sdplusmomentum1.pdf")
plt.show();
```

## Graficamos los gradientes

```python
plt.title("gradientes")
plt.plot(gradList, label="descenso con gradiente")
plt.plot(gradList2, label="descenso con gradiente + momentum")
plt.legend()
plt.savefig("gradient1.pdf")
plt.show()
```

## Momentum (version 3)
\begin{eqnarray}
v^{(k)} &=& \beta v^{(k-1)} + (1 - \beta) g^{(k)} \\
\theta^{(k)} &=& \theta^{(k-1} - \eta v^{(k)}.
\end{eqnarray}



```python
beta = 0.8
# initial point
x = 0.5

# list of points
v=0
xlist2 = []
ylist2 = []
gradList2=[]

for i in range(M):
    xlist2.append(x)
    ylist2.append(f(x))
    gradList2.append(gradf(x))
    g = gradf(x)
    v = beta*v + (1 - beta)*g
    x = x - eta*v
    # if abs(f(x)) < eps:
     #   break


xdomain = np.arange(-0.5, 0.51,0.01)
# cast lists into numpy array
xarr = np.array(xlist)
yarr = np.array(ylist)

xdomain = np.arange(-0.5, 0.51,0.01)
plt.plot(xdomain, f(xdomain));
# plt.plot(xlist,ylist, color='b');
plt.quiver(xarr[:-1], yarr[:-1], xarr[1:]-xarr[:-1], yarr[1:]-yarr[:-1], \
           scale_units='xy', angles='xy', scale=1, label="steepest descent", color="blue")

# cast list into np arrays
xarr2 = np.array(xlist2)
yarr2 = np.array(ylist2)
xdomain = np.arange(-0.5, 0.51,0.01)
plt.plot(xdomain, f(xdomain));
plt.plot(xlist,ylist, color='blue');
# plt.quiver(xlist2,ylist2, 0.1, 0.1, color='brown');
plt.quiver(xarr2[:-1], yarr2[:-1], xarr2[1:]-xarr2[:-1], yarr2[1:]-yarr2[:-1], scale_units='xy', \
           angles='xy', scale=1, label="steepest descent + momentum ", color="brown")
plt.scatter(xlist, ylist, s=80, facecolors='none', edgecolors='k');
plt.scatter(xlist2, ylist2, s=80, facecolors='none', edgecolors='r');

plt.legend()

plt.savefig("sdplusmomentum3.pdf")
plt.show();
```

# Metodo de aceleracion de Nesterov.
Se llama (NGA): Nesterov Gradient Accelereator. Yuii Nesterov, fue estudiante de doctorado de Poliak. El articulo del gradiente se publico en 1983 (referencia en las notas de clase).

La idea de nesterov era calcular el gradiente un paso mas adelante.
Es decir, el gradiente se calcula luego de que se avanza.
Se arranca de $\theta^{(i)}$, se avanza a $\theta^{(i)} + \beta v^{(i)}$
esto es momentum. Y el gradiente se evalua aca $g( \theta^{(i)} + \beta v^{(i)})$, mientras el momentum original lo evalue en $g(\theta^{(i)})$.

# AdaGrad (Adaptive)
Este metodo fue introducido por Duchi et. al. 2012 (referencia en las notas de clase). Este es el primero de los metodos que he expuesto donde $\eta$ es variable.
De esta forma la ecuacion de descenso es

$$\theta^{(i+1)} = \theta^{(i)} - \eta^{(i)} g^{(i)} .$$
Observe que aca $\eta$ puede cambiar en cada iteracion.
La idea es que el algoritmo se "adapta" a medida que avanzq.
Esta adaptacion viene de la idea de compensar el hecho de que el las componentes del gradiente sean muy distintas.

Agamos un ejemplo numerico
Considere la paraboloide eliptica

$$ f(x) = \frac12 x^T A x - b^T x + c   \quad , \quad x \in \mathbb{R}^n \quad , \quad A \in \mathbb{R}^{n \times n} \quad  , \quad c \in \mathbb{R}. \tag{1} $$
Asusimos que $A$ es positiva definida y simetrica.
Si aplicamos el teorema espectral,


$$ A = Q \Lambda Q^T.$$
donde $Q$ es ortogonal y $\Lambda$ es diagonal, con $\lambda_i > 0$.
Si reemplazamos esto en la Ecaucion (1), encontramos que si
$ x'  = Q^T x$, $x = Q x'$, i$x'^T = x^T Q$,

$$ f(x') = \frac12 x^T Q \Lambda  Q^T x - b^T x = \frac12 x'^T \Lambda x' - b^T Q x' + c.$$
Si calculamos el gradiente
$$ \nabla f(x') = \Lambda x' - Q^T b.$$
Las entradas de $\Lambda$ son todas positivas y

$$ \kappa = \frac{\lambda_1}{\lambda_n} .$$
es el numero de condicion ($\lambda_1 \ge \lambda_2 \ge \cdots \lambda_n $)
Si el numero de condicion fuera $1$ la superficie fuera una esfera.
El ejemplo numerico es el siguiente


$$f(x_1, x_2) = \frac{x_1^2}{9} + x_2^2 = \frac12 \left ( \frac{2 x_1^2}{9} + 2 x_2^2 \right ) = \frac12 (x_1   \quad x_2) \begin{pmatrix} 2/9 & 0 \\ 0 & 2\end{pmatrix}  \binom{x_1}{x_2} .$$
`













image.png

En este caso
$$\nabla f(x_1, x_2) = 2 \left ( \frac{x_1}{9} \quad x_2   \right ) .$$
Para el punto inicial $(-2, 2.2)$ tenemos

$$ \nabla f(-2, 2.2) = (-0.444, 4.4). $$
En la direccion $x_2$ el gradiente es 10 mas grande en tamano que en la direccion $x_1$ .

Todo esto fue motivacion para introducir el metodo de AdaGrad.

En el metodo de AdaGrad, el "learning parameter" esta dado por

$$\eta^{(i)} = \frac{\eta}{\sqrt{G^{(i)} + \epsilon}} . \tag{2}$$
El $\epsilon > 0$ es para evitar division por cero o valores muy pequenhos.

Quien es $G^{(i)}$?

Es suma (acumulado sobre la historia del desplazaminto)
$$G^{(i)} = \sum_{j=0}^i g^{(j)} \circ g^{(j)}.$$
donde $\circ$ es producto Hadamardi, y  $g^{(j)}$
es un vector, donde las componentes corresonden a la
componente $k$ del gradiente, para cada iteracion.
Mas claramente

Si
$$ g_k = (g_k^{(1)}, g_k^{(2)}, \cdots, g_k^{(i)}) \quad k=1,2, \cdots, n .$$
y
$$  g^{(j)} = (g_1^{(j)}, g_2^{(j)}, \cdots , g_n^{(j)} ) \quad , \quad  j=1,2, \cdots , i  ,$$

el producto hadamard
$$g^{(k)} \circ g^{(k)} =  ((g_1^{(k)})^2, (g_2^{(k)})^2, \cdots, (g_n^{(k)})^2) \quad k=1,2, \cdots, n .$$
De forma que

$$G^{(i)} = ( \| g_1  \|^2, \| g_2 \|^2 , \cdots \| g_n \|^2)  .$$
La historia acumulada de cada componente del gradiente.

En la formula (2) parece haber un error de dimensiones por que se divide por un vector. Se puede dividir por un vector?, no
Entonces la formula adecuada es


$$\eta_k^{(i)} = \frac{\eta}{\sqrt{G_k^{(i)} + \epsilon}} . \tag{2}$$

Lo que realmente hace el algoritmo es:


$$\theta^{(i)} = \theta^{(i-1)} - \eta^{(i)} \circ g^{(i)} .$$
Tenemos al final, de nuevo, un producto Hadamard, que se hace
componente a componente.
