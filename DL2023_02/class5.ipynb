# Deltas de Kronecker


\begin{eqnarray}
\delta_i^j = \left \{
    \begin{array}{cc}
    1  & \text{if } i = j \\
    0 & \text{if } i \ne j
    \end{array}
    \right .
\end{eqnarray}

Se usan mucho cuando tenemos derivadas de coordenadas mutuamente ortogonales. Por ejemplo en el plano Cartesiano $\in \mathbb{R}^2$, coordenadas $(x_1, x_2)$. Tenemos la siguiente propiedad


\begin{eqnarray}
\frac{\partial x_j}{\partial x_i} = \delta_i^j.
\end{eqnarray}

Como matriz $\delta_i^j$ es la matriz identidad. Hay propiedades muy importantes.


$$ \delta_i^j x_j = x_i .$$

Las identidades $\epsilon-\delta$.

La primera


\begin{eqnarray*}
  \left |
  \begin{array}{ccc}
    \delta_m^r & \delta_n^r & \delta_p^r \\
    \delta_m^s & \delta_n^s & \delta_p^s \\
    \delta_m^t & \delta_n^t & \delta_p^t
  \end{array}
  \right |
  &=& \epsilon^{rst} \epsilon_{mnp} ,
\end{eqnarray*}

Expandiendo el determinante encontramos:

\begin{eqnarray*}
  \left |
  \begin{array}{ccc}
    \delta_m^r & \delta_n^r & \delta_p^r \\
    \delta_m^s & \delta_n^s & \delta_p^s \\
    \delta_m^t & \delta_n^t & \delta_p^t
  \end{array}
  \right |
  &=&  \delta_m^r \delta_n^s \delta_p^t + \delta_n^r \delta_p^s \delta_m^t + \delta_p^r \delta_m^s \delta_n^t - \delta_m^t \delta_n^s \delta_p^r - \delta_n^t \delta_p^s \delta_m^r - \delta_p^t \delta_m^s \delta_n^r .
\end{eqnarray*}

Llamemos $a_{mnp}^{rst}$ a la expresion anterior, pero los voy a conmutar de forma que todos los subindices esten en el orden $m,n,p$.

$$ a_{mnp}^{rst} = \delta_m^r \delta_n^s \delta_p^t +  \delta_m^t \delta_n^r \delta_p^s + \delta_m^s \delta_n^t \delta_p^r - \delta_m^t \delta_n^s \delta_p^r - \delta_m^r \delta_n^t \delta_p^s - \delta_m^s \delta_n^r \delta_p^t \tag{1} $$

Vamos a probar que $a_{mnp}^{rst}$ es totalmente anti-simetrico en los indices $r,s,t$. Es decir que al permutar cualquier de esos indices, cambia el signo, pero no su valor absoluto.

Veamos el caso para $r,s$ (deamos los otros dos $r,t$, $s,t$) al estudiante.

En la expresion cambiemos todos los $r$ por $s$ y los $s$ por $r$



\begin{eqnarray}
 a_{mnp}^{srt} &=& \delta_m^s \delta_n^r \delta_p^t +  \delta_m^t \delta_n^s \delta_p^r + \delta_m^r \delta_n^t \delta_p^s - \delta_m^t \delta_n^r \delta_p^s - \delta_m^s \delta_n^t \delta_p^r - \delta_m^r \delta_n^s \delta_p^t \\
&=& -\delta_m^r \delta_n^s \delta_p^t - \delta_m^t \delta_n^r \delta_p^s - \delta_m^s \delta_n^t \delta_p^r + \delta_m^t \delta_n^s \delta_p^r + \delta_m^r \delta_n^t \delta_p^s + \delta_m^s \delta_n^r \delta_p^t \\
&=& -a_{mnp}^{rst}.
\end{eqnarray}
Vemos que al reversar $r$ con $s$, el signo de $a_{mnp}^{rts}$ cambia. Se deja como prueba los otros casos expuestos arriba.

Es decir, se deja como prueba

$$ a_{mnp}^{tsr} = -a_{mnp}^{rst} \quad , \quad a_{mnp}^{rst} = - a_{mnp}^{rst} $$

Entonces el tensor $a_{mnt}^{rst}$ es totalmente antisimetrico en los indices $r,s,t$ y por lo tanto usando una expresion de la clase anterior

$$ a_{mnp}^{rst} = \epsilon^{rst} a_{mnp}^{123} \tag{2}$$

Con,

$$a_{mnp}^{123} =
\delta_m^1 \delta_n^2 \delta_p^3 +  \delta_m^3 \delta_n^1 \delta_p^2 + \delta_m^2 \delta_n^3 \delta_p^1 - \delta_m^3 \delta_n^2 \delta_p^1 - \delta_m^1 \delta_n^3 \delta_p^2 - \delta_m^2 \delta_n^1 \delta_p^3  $$

Observamos que:
Si $(m \; n \; p)$ es una permutacion par de $(1 \; 2 \3)$ entonces
$a_{mnp}^{123}$ produce $1$. Si $(m \; n \; p)$ es una permutacion impar de $(1 \; 2 \;3)$ entonces $a_{mnp}^{123}=-1$.
Si dos de los indices son repetidos, el resultado es 0.
Esto lo dejamos como prueba al estudiante.  Entonces
$a_{mnp}^{123}$ es un Levi-Civita. O sea

$a_{mnp}=\epsilon_{mnp}$ y encontramos que


$$ a_{mnp}^{rst} = \epsilon^{rst} \epsilon_{mnp} \tag{2}$$


Renombramos la expresion anterior como otro Delta de Kronecker:

$$ \delta_{mnp}^{rst} = \epsilon_{mnp} \epsilon^{rst}. \tag{3}$$

Este delta de Kronecker tiene la siguiente propiedad.

\begin{eqnarray}
\delta_{mnp}^{rst} = \left \{
    \begin{array}{cc}
    0 & \text{ si dos indices entre $r,s,t$ son repetidos} \\
    1 & \text{ si $(r \; s \; t)$ es una permutacion par de $(m \; n \; p)$} \\
    -1 & \text{ si $(r \; s \; t)$ es una permutacion impar de $(m \; n \; p)$} \\
    \end{array}
    \right .
\end{eqnarray}

Veamos conceptos como : Gradiente, divergencia, Laplaciano, Jacobiano,

$a=(a_1, a_2, \cdots, a_m)$.

* **Operador coma**

* Gradiente:
$$ \nabla a :   a_{,i}$$

El simbolo $_,i$  indica derivada parcial con respecto a $x_i$
Es decir

$$ a_{,i} = \frac{\partial a}{\partial x_i}. $$

$$\nabla^2 a : a_{,ii}. $$

$$ a_{,ii} = \sum_{i=1}^n \frac{\partial^2 a}{\partial x_i^2} $$

$$ \nabla \cdot a:  a_{i,i}.$$

**El Jacobiano.**
Asumimos que $y : \mathbb{R}^n \to \mathbb{R}^n$.


$$ J = y_{,j}^i$$

\begin{eqnarray*}
  J = {y^i}_{,j} \quad : \quad
  \begin{pmatrix}
    \frac{\partial y^1}{\partial x_1} & \frac{\partial y^1}{\partial x_2} & \cdots
    & \frac{\partial y^1}{\partial x_n} \\
    \frac{\partial y^2}{\partial x_1} & \frac{\partial y^2}{\partial x_2} & \cdots
    & \frac{\partial y^2}{\partial x_n} \\
    \vdots & \vdots & \ddots & \vdots  \\
    \frac{\partial y^n}{\partial x_1} & \frac{\partial y^n}{\partial x_2} &  \cdots
    & \frac{\partial y^n}{\partial x_n} \\
  \end{pmatrix} .
\end{eqnarray*}




Esta tecnologia es muy util en problemas de optimizacion.
Piense en una funcion objetivo (cuadratica) como


$$J(x)= \frac12 x^T A x - x^T b + c $$
donde $x,b,c \in \mathbb{R}^n$, $A$ es una matrix $n \times n$.

Se quiere sacar el **gradiente** de $J$ para obtimizar.

Usando notacion indicial encontramos.
Asumimos que la matriz $A$ es constante.
Asumimos que $A$ es simetrica


\begin{eqnarray}
J_{,k} &=& \left [ \frac12 a^{ij} x_i x_j - b^i x_i + c^i
\right ]_{k} \\
&=& \frac12 a^{ij} [ x_i x_{j_{,k}} + x_{i_{,k}} x_j] - b^i x_{i_{,k}} \\
&=& \frac12 a^{ij} [ x_i \delta_j^k + \delta_i^k x_j] - b^i \delta_i^k \\
&=& \frac12 a^{ik} x_i + \frac12  a^{kj} x_j - b^k \\
&=& \frac12 a^{ik} x_i + \frac12  a^{ki} x_i - b^k \\
&=& \frac12 a^{ik} x_i + \frac12  a^{ki} x_i - b^k \\
&=& a^{kj} x_j - b^k.
\end{eqnarray}
En forma de bloques esto es

$$ \nabla J(x) = Ax - b $$

De la misma forma que si $f(x)=\frac12 a x^2 - bx + c $ (en los reales $f'(x) = ax -b$.

Si $A$ no es simetrica lo mira en las notas.
