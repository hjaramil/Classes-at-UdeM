# Metodo regular de descenso de gradiente
De Calculo en varias variables, la direccion de maximo descenso es

$$-\frac{\nabla_{\theta} C}{ \| \nabla_{\theta} C \|} .$$
donde $C(\theta)$ es una funcion de costo en el parametro $\theta$, $\theta \in \mathbb{R}^n$.


Como descendemos.
$$ \theta^{(i+1)} = \theta^{(i)} - \eta \nabla C(\theta^{(i)}) \tag{1} .$$
Expliqemos esta ecuacion.
El nuevo parametro (esta en la abscisa de $C(\theta)$, o sea que es el nueva $\theta$. ) apartir de $\theta^{(i)}$ es $\theta^{(i+1)}$, el signo $-$ es por que queremos descender, no ascender, es la direccion negativa del gradiente. $\nabla C(\theta^{(i)})$ es el gradiente en el punto anterior (ver dibujos  mano alzada).
No esta normalizado por que este factor $1/\| \nabla_{\theta} C \|$ lo consideramos absorvido por el parametro $\eta$ que se llama el **learning rate** (rata de aprendizaje).

Cuando se usa la Ecuacion (1) sin mas preambulos el metodo se llama Steepest Descent (maximo descenso). Hay direcciones mas rapidas que el maximo descenso (esto es aparentemente contradictorio). Veamos un ejemplo. Ver notas a mano alzado.






[Gradiente Conjugado](https://en.wikipedia.org/wiki/Conjugate_gradient_method).
Hestenes, Stephen, Lanczos lo descubrieron el mismo anho.
En un plano va directo al centro de la elipse, tomando dos pasos, el inicial y el siguiente. En el espacio de 3D toma 3 pasos (maximo) para llegar al minimo. En $n$ dimensiones toma maximo $n$ iteraciones.

## Stochastic Gradient Descent.
Metodos de descenso de gradiente estocasticos.
Este metodo se introdujo en 1951 por Robbins y Monro.
(Herbert Robbins y Sutton Monro)


### Motivacion.
Observe la linea roja en el algortimo. El calculo de este gradiente puede ser muy costoso.

Ejemplo con redes neuronales.

Asuma una red neuronal (no profunda,
* por que solo la asumimos con 9 capas ocultas y de salida. (1 de entrada y otras 10).
* La entrada para su entrenamiento requiere 10 mil imagenes. Cada imagen tiene 100x100 pixeles (en gris, si es a color toca multiplicar por 3 o por 4).
* Cada capa despues de la entrada tiene 100 neuronas.

Puntos de entrada (neuronas de entrada). $1000 \times 100 \times 100 = 10^8 $
El numero de pesos es.
$$10^8 \times 100^{10} = 10^{8+20}= 10^{28}. $$

Este ejemplo no incluye dimensiones desorbitantes, sin embargo el numero de pesos no cabe en un computador moderno. Este problema
se debe reducir. El gradiente se calcula para cada una de estas componentes. El gradiente tendria la dimension de $10^{28}$.
Tampoco cabe en ningun computador.


Hay dos cosas a considerar:

* No usar todos los pesos. Esta es la idea central de CNN: Convolutional Neural Networks
* No calcule todas las componentes del gradiente. Aca es donde aparece el metodo de SGD.
