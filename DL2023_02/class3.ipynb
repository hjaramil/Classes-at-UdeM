Para nosotros (en ciencia de la computacion) un tensor simplemente es un objeto multi-indice. El rango de un tensor es el numero de indices libres. Por ejemplo

$$ a_{ijk}^{k} $$
El indice $k$ es "dummy" (mudo) no se cuenta. Entonces el rango de este tensor es $2$, solo contamos los libres que son $i,j$.

## Algebra de Tensores:
Vamos a hablar de suma-restas y productos. Los productos son de distintas clases, "outer", "inner"-> contraccion.

* Suma-resta:  Dados dos tensores
$ a_{i_1 i_2 \cdots, i_n}^{j_1 j_2 \cdots j_n} \quad $  ,
$ b_{i_1 i_2 \cdots, i_n}^{j_1 j_2 \cdots j_n}$.
Definimos la suma/resta como
\begin{eqnarray}
 c_{i_1 i_2 \cdots, i_n}^{j_1 j_2 \cdots j_n} = a_{i_1 i_2 \cdots, i_n}^{j_1 j_2 \cdots j_n} \pm b_{i_1 i_2 \cdots, i_n}^{j_1 j_2 \cdots j_n}
\end{eqnarray}
Note que las dimensiones deben coincidir en ambos tensores.

* Producto tensorial: Dados dos tensores
$ a_{i_1 i_2 \cdots, i_n}^{j_1 j_2 \cdots j_n} \quad $  ,
$ b_{k_1 k_2 \cdots, k_p}^{l_1 l_2 \cdots l_q}$.
Definimos el producto tensorial (outer=externo) como

$$c_{i_1 i_2 \cdots, i_n k_1 k_2 \cdots  k_p}^{j_1 j_2 \cdots j_n l_1 l_2 \cdots l_q} =  a_{i_1 i_2 \cdots, i_n}^{j_1 j_2 \cdots j_n} \otimes b_{k_1 k_2 \cdots, k_p}^{l_1 l_2 \cdots l_q} =
 a_{i_1 i_2 \cdots, i_n}^{j_1 j_2 \cdots j_n}  b_{k_1 k_2 \cdots, k_p}^{l_1 l_2 \cdots l_q} $$

Estamos asumiento que no hay indice comun.

Que pasa si hay un indice comun? se suma a lo largo de el con la notacion de Einstein.

De aca viene la palabra **contraccion**.

* Contraccion: Si en un tensor se hace un indice covariante igual al contravariante se suma sobre ese indice. Ejemplo:


$$ a_{i_1 \cdots m \cdots i_n}^{j_1 \cdots m \cdots j_l}
=\sum_m a_{i_1 \cdots m \cdots i_n}^{j_1 \cdots m \cdots j_l} $$
donde escribo el sumbolo $\sum$ para enfatizar que se suma sobre el indice mudo.  Se puede contraer en 2 el rango. Esto da origen
al producto interno

* **Producto Interno**: Aca se asume que los tensores comparten algunos indices y esto da origen a sumas sobre estos indices.

Ejemplos:

** $a_i^j x_j $: Multiplicacion de la matriz $A$ por el vector $x$.  
** $x_i y^i$ : Este es un escalar. Y el producto punto como lo concemos.

** $a_{ijkl}^{pq} b_{pq}  $: Un tensore de orden 4. en los indices $i,j,k,l$.





## Tensores simetricos y antisimetricos (skew-symmetric)
Dado un sistema con dos indices $a_{ij}$ decimos que es **simetrico** si $a_{ij}= a_{ji}$. En general para sistemas de cualquier rango, decimos que el sistema es simetrico en dos de sus indices , si al reversar el orden en los indices no hay cambios. Por ejemplo

$a_{i_1 i_2 \cdots i_j \cdots i_k \cdots i_n}
= a_{i_1 i_2 \cdots i_k \cdots i_j \cdots i_n} $
Este tensor es simetrico en los indices $i_j$ , $i_k$

Ahora, si es simetrico en todos sus indices se dice que es
**completamente simetrico**.

Por ejemplo tomemos el tensor de rango 3 $a_{ijk}$. Si es completamente simetrico tenemos que

$$ a_{ijk} = a_{jki} = a_{kij} = a_{kji} = a_{jik} = a_{ikj} $$
Este tensor es totalmente $a_{123} \alpha$, donde $\alpha \in \mathbb{R}$.

Ejemplos:

* $a_{ijk} = i + j - k$, Es simetrico en $i, j$.
* $a_{ijk} = i + j + k$, Es simetrico en $i, j,k$.  Es completamente simetrico (por la ley conmutativa de la suma).

* $a_{ijk} = ij + k$. Que simetria se observa aca?


### Tensor antisimetrico (skew-symmetric)
Un sistema es antisimetrico en dos de sus indices si al intercambiarlos, cambia el signo de la componente correspondiente.

Por ejemplo:


$a_{i_1 i_2 \cdots i_j \cdots i_k \cdots i_n}
= -a_{i_1 i_2 \cdots i_k \cdots i_j \cdots i_n} $
Aca intercambiamos la posicion del indice $i_j$ con el $i_k$ y el signo se reverso.

Si a este tensor antisimetrico en los indices $i_j$, $i_k$ hacemos los indices iguales $i_j=i_k$ el valor es cero. Por que tendriamos un elemento igual a su negativo y solo el cero cumple esta propiedad. O sea que si $a_{i_1 i_2 \cdots i_j \cdots i_k \cdots i_n} $ es antisimetrico en $i_j$ e $i_k$ y por ejemplo
$i_j=i_k=3$ entonces

$$ a_{i_1 i_2 \cdots 3 \cdots 3 \cdots i_n} = 0 $$

Si el tensor es antisimetrico en **todos** sus indices se dice que es completamente antisimetrico.

Ejemplos:

* $a_{ij} = i - j$, Es completamente antisimetrico. Ejercicio: pruebelo

* $a_{ijk} = i + j - k$ es simetrico? antisimetrico? discuta.



Existen los tensores de orden 3 completamente antisimetricos?
Si.


Levi-Civita (Italiano) definio el tensor de orden 3 totalmente antisimetrico.

\begin{eqnarray}
\epsilon_{ijk} = \left \{
    \begin{array}{cc}
    1  & \text{si   } \quad  (i,j,k) \in \{ (1,2,3), (2,3,1), (3,1,2)\} \\
    -1  & \text{si   } \quad  (i,j,k) \in \{ (1,3,2), (3,2,1), (2,3,1)\} \\
    0  & \text{en los demas casos, cuando hay dos o tres repetidos}
    \end{array}
    \right .
\end{eqnarray}

Si un tensor de rango 3 es totalmente antisimetrico entonces
es de la forma

$$ a_{ijk} = \epsilon_{ijk} a_{123} $$
