Se explico una primera aproximacion de ANN: Artificial Neural Network  a mano alzado. Un modelo de aprendizaje profundo requiere muchas capas (~50 para arriba). 3 capas no dan para deep learning. A 3 capas se le llama MLP: Multilayer Perceptron. Esto lo veremos mas en detalle en el curso de ML.


Respondiendo a su pregunta:

ML tiene tres areas

* Aprendizaje Supervisado: Tiene etiquetas. Al ninho se le dice que la figura es de un elefante. La palabra "elefante" es la etiqueta.
* Apendizaje no supervisado: Al ninho no  se le dice la palabra.
Se le entregan 100 cartas. El ninho las clasifica. Sin saber que es un elefante, coloca todos elefantes en un monton, las girafas en otro y asi suscesivamente. A esto se le llama "clustering".
* Aprendizaje re-enforzado (reinforcement): Se usa para juegos.
Existe un agente (el jugador) y entonces existe un ambiente. El jugador no tiene idea de como jugar. Pero a medida que juega gana y pierde. Al principio pierde mucho, y cuando pierde tiene penalties y cuando gana recompenzas, entonces va a aprendiendo por que cuando falla sabe no debe volver a repetir esta jugada y asi.

Tambien existe un metodo llamado semi-supervisado. Por ejemplo en una tecnica que se llama "sistemas de recomendacion", por ejemplo Neflix. Algunos usuarios califican la pelicula con 1,2,3,4,5 estrellas otros no. La calificacion es la respuesta de la figura 1. Otros no. Los que califican "supervisan" los que no "no supervisan" y el sistema es semi-supervisado.

Por ejemplo en YouTube. Usted no califica las canciones que escucha, pero la maquina lee de las veces que la escuchado o los estilos que ha escuchado toma esos datos y los usa para sugerir o tocar la proximas cancion sin usted la seleccione.

Dentro de Aprendizaje supervisado:

* Regresion lineal
* Regresion logistica
* SVM: Support Vector Machines
* ANN

Dentro ANN esta DL (cuando las capas son muchas).

DS-> AI->ML->supervisado-> ANN->Dl

Ojo!!!! puede haber ANN no supervisado o semi-supervisado
y entonces DL puede ser supervisado, no supervisado o incluso
re-enforzado.

No hay un "sharp boundary" entre DL y los distintos tipos de aprendizaje. EN DL pueden ocurrir aprendizajes supervisados, o no supervisados, o de refuerzo (reinforcement).

## Historia:

La historia de ML la veremos en el curso de ML.

Aca solo en cosas concretas del curso que vamos a ver


* 1960: Henry Kelly: desarrollo el algoritmo de **back propagation** (lo explicare en de detalle en el curso de ML). Tambien aca en 1962 Stuart Dreyfus desarollo una version mas simple usando la regla de la cadena.  Esta tecnica evoluciono mucho En 1970 Seppo Linnainmaa desarrolo el algoritmo en FORTRAN  para su tesis de maestria. En 1985 Rumhelhart Williams y Jeffrie Hinton desarrollaron una version de BP mas practica. Pero fue Yann LeCun en 1989 en Laboratorios Bell hizo la primera implementacion mass practica con resultados que veremos en el curso.


* 1979: Kunihiko Fukushima invento las CNN: Convolutional Neural Networs (redes neuronales convolucionales). Estas las vamos a ver en el curso con mucho detalle.

* 1995: Se introdujo las LSTM: Long Short Term Memory (memoria corta y larga). Son muy utiles para el NPL: Natural Language Processing.

* 1999: Se descubre el problema del "vanishing gradient" (gradiente que desvanece).

* 2009: Profesora Fei-Fei Li (Stanford). Lanzo el ImageNet que tiene mas 15 millones de imagenes con etiquetas (labels). Ha servido para concursos gigantes como el ILSVRC: ImageNet Large Scale Visual Recognition Challenge. Este concurso ha impulsado la intelegencia artificial de una forma jigantesca. Por ejemplo el AlexNet (lo vamos a estudiar) disenhado por Alex Krizhevsky (el grupo de Hinton) gano premio y pudo impulsar el desarrollo del la perte visual (imaging Visual Computation). Vamos a ver esto en el curso. Uso por primera ves GPU (Graphics Processing Units) en este proyecto.

* 2010. Hay otra institucion dedicada a los concursos con DL. Se llama Kaggle. Ha impulsado grandemente el desarrollo de DL.
