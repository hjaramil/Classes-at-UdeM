# The gradient (Jacobian) of a tensor
De la forma mas general asumamos que $\tau$ es un tensor
$\tau = \tau_{i_1 i_2 \cdots i_p} $ es un tensor de rango $p$.
con $i_r = 1,2, \cdots, I_r$, $r=1,2, \cdots, p$ y si
$\Theta$ es un tensor $\Theta=\Theta_{j_1 j_2 \cdots j_q}$ de rango $q$, con $j_s = 1, 2 \cdots, J_s$, $s=1,2 \cdots, q$.
Entonces:


$$ J = \nabla_{\Theta} \tau = \frac{\partial \tau}{\partial \Theta}= \begin{pmatrix} \frac{\partial \tau_{i_1 i_2 \cdots i_r} }{\partial \theta_{j_1 j_2 \cdots j_s} } \end{pmatrix} .$$

Este seria un tensor (Jacobiano) de rango $p+q$ en

$$\mathbb{R}^{(I_1 \times I_2 \cdots \times I_p) \times (J_1 \times J_2 \times \cdots \times J_s)}.
$$

Desde el punto de vista de algebra lineal este espacio tiene dimensiones $\prod_{i=1}^p I_r \prod_{j=1}^q J_j$.
El estudiante puede verificar que todos casos anteriores son casos particulares de esta ecuacion general. Por ejemplo si $f=\tau$ es escalar (rango cero) y $\Theta$ una matriz (rango 2) entonces $\nabla_{\Theta} \tau$  es un tensor de rango 2 con $p=0, q=2, I_0=0, J_1=m, J_2=n$, una matriz en $\mathbb{R}^{m \times n}$ .

## La regla de la cadena:
En la forma mas simple si $f, g$ son funciones reales y
$h(x)=(f \circ g)(x) = f(g(x)) $, entonces la derivada esta dada por

$$h'(x)= f'(g(x)) g'(x) = \frac{dh}{dx} = \frac{d f}{dg} \frac{dg}{dx} .$$
Extendamos un poco. Asuma que $f$ es un campo escalar con $x \in \mathbb{R}^n$ y que $g=g(t)$ es una funcion vectorial $(g_1(t), g_2(t), \cdots, g_n(t))$, entonces


$$\frac{dh}{dt} = \sum_{i=1}^n \frac{\partial f}{\partial g_i} \frac{d g_i}{d t} = \frac{\partial f}{\partial g} \frac{dg}{dt} .$$

donde la ultima operacion es el producto de un vector fila con un vector columna , o mejor un producto punto, o interno.

Generalizamos un poco mas. Asumamos

\begin{eqnarray}
f : A \subset \mathbb{R}^p &\to& \mathbb{R}^q \\
g : B \subset \mathbb{R}^m &\to& \mathbb{R}^p \\
h : B \subset \mathbb{R}^m &\to& \mathbb{R}^q \\
x &=& \mapsto& h(x) = (f \circ g)(x)  = f(g(x))
\end{eqnarray}
La regla de la cadena para este caso se traduce en:


$$\frac{\partial h_i}{\partial x_j} = \sum_{k=1}^p \frac{\partial f_i}{\partial g_k} \frac{\partial g_k}{\partial x_j} \in \mathbb{R}^{q \times m} \quad, i=1, \cdots, q \quad , \quad j=1, \cdots, m .$$

Desde el punto de vista matricial tenemos






$$\frac{\partial h}{\partial x} = \frac{\partial f}{\partial g} \frac{\partial g}{\partial x} $$
donde $\partial f/\partial g \in \mathbb{R}^{p \times q}$ y $\partial g/\partial x \in \mathbb{R}^{p \times m}$.

## Caso especial usado en LSTM (Long Short Time Memory).

Sea $x \in \mathbb{R}^{n_0}$, $f_1(x) \in \mathbb{R}^{n_1}$, $\cdots f_k(x) \in \mathbb{R}^{n_k}$  definimos
$g(x)= f_k(f_{k-1} ( \cdots f_1(x) \cdots )  $
Entonces, escribiendo en notacion matricial


$$ \frac{\partial g}{\partial x} = \frac{\partial f_k}{\partial f_{k-1}} \frac{\partial f_{k-1}}{\partial f_{k-2}} \cdots \frac{\partial f_2}{\partial f_1} \frac{\partial f_1}{\partial x} = \in \mathbb{R}^{n_k \times n_0} .$$

## Hadamard Products (Productos de Hadamard)
Asuma dos campos vectoriales $f=(f_1, f_2, \cdots, f_n)^T$ ,
$g=(g_1, g_2, \cdots, g_n)^T$ ambos en $\mathbb{R}^n$, de variable
$x \in \mathbb{R}^m$.  El producto Hadamard esta definido por


$$ h = f \circ g  = (f_1 g_1, f_2 g_2, \cdots, f_n g_n)^T .$$
donde $\circ$ significa aca producto Hadamard (no composicion de funciones).

Entonces la derivada del producto Hadamard aca es


\begin{eqnarray}
\frac{\partial h }{\partial x_i}
= \begin{pmatrix}
\frac{\partial f_1}{\partial x_i} g_1 + f_1 \frac{\partial g_1}{\partial x_i} \\
\frac{\partial f_2}{\partial x_i} g_2 + f_2 \frac{\partial g_2}{\partial x_i} \\
\vdots &  \\
\frac{\partial f_n}{\partial x_i} g_n + f_n \frac{\partial g_n}{\partial x_i}
 \end{pmatrix}
 = \frac{\partial f}{\partial x_i} \circ g + f \circ \frac{\partial g}{\partial x_i}
\end{eqnarray}

Podemos generalizar la expresion como
$$\left ( \frac{\partial h}{\partial x} \right )_i = \left ( \frac{\partial f}{\partial x} \right )_i
\circ g  + f \circ \left ( \frac{\partial g}{\partial x} \right)_i $$


Un caso particular que lo usamos en el desarrollo de las RNN.

Asuma que $x=f$ entonces

$$\frac{\partial h}{\partial x}  =  \frac{\partial f}{\partial f}
\circ g  + f \circ \frac{\partial g}{\partial x_i} =  I \circ g + f \circ \frac{\partial g}{\partial x}$$
donde $I$ es la matriz identidad en $\mathbb{R}^n$ y
$I \circ g$ es la matriz diagonal con elementos $g_i$ a lo largo de la diagonal. Veamos esto de forma explicita.

$$\frac{\partial h_i}{\partial x_j} = \frac{\partial x_i g_i}{\partial x_j} =
\frac{\partial x_i}{\partial x_j} g_i + x_i \frac{\partial g_i}{\partial x_j} = \delta_{ij} g_i + f_i \frac{\partial g_i}{\partial x_j} .$$




# Gradiente en una RNN:
image.png

La funcion de costo global es:


$$ C = \frac{1}{T} \sum_{t=1}^T C_t.$$
Al gradiente de esta funcion de costo con respecto a los $\Theta$s se le llama gradiente de
[BPTT](https://en.wikipedia.org/wiki/Backpropagation_through_time).
Back Propagation Through Time.

En esta clase no alcanzamos a calcular el gradiente de $C$ pero vamos iniciar el proceso y lo terminamos en la proxima clase.

Para simplificar solo tomemos un $C_k$ (al final el gradiente es la suma de todos ellos


$$ \nabla_{\Theta} C  = \frac{1}{T}  \sum_{k=1}^T \nabla_{\Theta} C_k .$$

Nos enfocamos en calcular

$$ \nabla_{\Theta} C_k = \frac{\partial C_k}{\partial \Theta} \tag{1}.$$

Formulemos mas claramente las funciones que estan en fuego.

$$ h_t = f(x_t, h_{t-1}, \theta_x, \theta_h).$$
$$ y_k = g(h_t, \theta_y) .$$

La pregunta es como evaluamos la eucuacion (1) (por ejemplo con respecto a $\theta_h$).  Obvio que tambien necesitamos las derivadas parciales con respecto a $\theta_x, \Theta_y$ pero
la mas dificil es con respecto a $\theta_h$, pues es recursiva.

Regla de la cadena.

$$\frac{\partial C_k}{\partial \theta_h} = \frac{\partial C_k}{\partial \hat{y}_{t}} \frac{\partial \hat{y}}{\partial h_t} \frac{\partial h_t}{\partial \theta_h}.$$

Tambien con respecto a $\theta_x$ y a $\theta_y$. Esto viene despues.

Por ahora lo dejamos aca y en la proxima clase, terminamos la evalucion del gradiente en un red neuronal RNN.
