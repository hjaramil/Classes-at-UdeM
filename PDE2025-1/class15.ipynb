Considere el siguiente IBVP: Initial Boundary Value Problem

$$
\begin{cases}
u_{tt} - c^2 u_{xx}  = F(x,t) \quad , \quad -\ell < x < \ell \quad c=\text{const} \\
u(x,0) = \varphi(x)\quad  , \quad u_t(x,0) = \psi(x)  \quad , \quad \text{condiciones iniciales} \\
u(\pm \ell, t) = h_{\pm}(t) \quad , \quad \quad h_{\pm}(0)=\phi(\pm \ell) \quad , \quad \frac{\partial h_{\pm}}{\partial t}(0) = \psi(\pm \ell)
\end{cases}
$$
Note que las dos ultimas condiciones parecen redundantes pero es para garantizar consistencia en los extremos.

$$\int_{PA} u_t dx + c^2 u_x dt = \int_{PA} c u_t dt + c u_x dx = c \int_{PA} du = c (u(A) - u(P) )$$

Se hace lo mismo con los demas.

$$\int_{AB} u_t dx + c^2 u_x dt = c( u(A) - u(B)) \quad, \quad dx = -c dt $$

$$\int_{AB} u_t dx + c^2 u_x dt = c( u(C) - u(B)) \quad, \quad dx = c dt $$

$$\int_{AB} u_t dx + c^2 u_x dt = c( u(C) - u(P)) \quad, \quad dx = -c dt $$

Debemos despejar $u(P) pues los demas puntos tienen valores conocidos.

$$u(P)= u(A) + u(C) - u(B) + \frac{1}{2c} \iint_{PABC} F(x,t) dx dt $$

De aca sale una propiedad interesante llamada **propiedad del paralelogramo caracteristico**. No que podemos para el $-u(B)$ a la izquierda y escribir

$$u(P) + u(B) = u(A) + u(C)  + \frac{1}{2c} \iint_{PABC} F(x,t) dx dt $$

Antes de estudiar el metodo de separacion de variables vamos a estudiar el **Sturm Liouville problem**.

## El Sturm-Liouville problem: Una maquina para generar bases ortogonales.
En algebra lineal nosotros usamos matrices $A \in \mathbb{R}^{n \times n}$ (cuadradas). Definimos el **autovector** o **vector propio** $u$ con **autovalor** o **valor propio** ($\lambda$) a los objetos tales que

$$A u = \lambda u $$
No nos intersa que $u=0$, por obvio $A 0 = \lambda 0=0$. Asumimos que $u \ne 0$.

Volveremos sobre esto mas adelante. Vamos a definir ahora el Sturm-Liouville problem

**Definicion 4.1.2** (Sturm-Liouville ODE):  A la ecuacion ODE de segundo orden de la forma

$$L u = \frac{1}{w(x)} \frac{d}{dx} \left ( p(x) \frac{du}{dx} \right ) + q(x) u = \lambda u $$

donde el peso (weight) $w(x)>0$. $\lambda$ es constante. A esto se llama la ecuacion de Sturm-Liouville.

Vamos a asumir que $u$ esta definido en un dominio real $[a,b]$ con condiciones de frontera

$$\alpha_1 u(a) + \alpha_2 u'(a) = 0 \quad , \quad \beta_1 u(b) + \beta_2 u'(b) = 0 $$
donde $\alpha_1^2 + \alpha_2^2 > 0$ y $\beta_1^2 + \beta_2^2 > 0$ para evitar
la ecuacion $0=0$ que no brinda ninguna informacion. A estas condiciones de frontera se les conoce como **mixtas** o **Robin**.

De aca salen dos casos especiales

$$u(a) = u(b)=0 \quad, \quad \alpha_2 = \beta_2 = 0 \quad , \quad \text{Dirichlet} $$
$$u'(a) = u'(b)=0 \quad, \quad \alpha_1 = \beta_1 = 0 \quad , \quad \text{Neumann} $$

La ecuacion de Sturm-Liouville se origino en 1836 por los autores Charles Francois Sturm y Joseph Liouville. (la referencia en las notas de clase).
Los problemas relacionados al operador $L$ de arriba se llaman
**Sturm-Liouville Problems** (SLP).

Recordemos el operador
$$Lu=\lambda u$$
que es un problema autovalor/autovector.  Pero en este caso (continuo) de espacios vectoriales infinito-dimensionales (de funciones) se llaman
**autovalores** (o **valores propios**) y los $u$ son **autofunciones** o **funciones propias**.

$$L u = \frac{1}{w(x)} \frac{d}{dx} \left ( p(x) \frac{du}{dx} \right ) + q(x) u = \lambda u \quad , \quad a < x < b$$
Asumimos que $u \ne 0$.

**Ejemplo 4.1.1**: Resuelva el SLP:

$$-u''(x) = \lambda u(x) \quad , \quad 0 < x < \pi $$
$$u(0) = 0 \quad , \quad u(\pi) = 0. $$

Encontramos que

\begin{eqnarray}
\text{para } \lambda &=& 1 \implies  u=u_1(x) = \sin x \\
\text{para } \lambda &=& 4 \implies  u=u_2(x) = \sin 2 x \\
&\vdots& \\
\text{para } \lambda &=& n^2 \implies u = u_n(x) = \sin n x.
\end{eqnarray}

Por que nosotros escogimos los $\lambda$ enteros?  Veamos que no nos queda de otra

1. Si $\lambda=0$. Es decir $Lu=0$, $-u''(x)=0$ e integramos
$$ u(x) = A x + B  \quad , \quad A, B \quad \text{   constantes} $$
$A,B$ se hallan con las condicines de frontera

$$u(0)=0 \implies A0 + B = 0 \implies B=0 $$
O sea que $u(x)=Ax$.
Pero como $u(\pi)=0$, entonces $A \pi = 0$, o sea que $A=0$.
De forma que la solucion es
$$ u(x) = 0$$
pero esta no es de interes. Entonces debemos escoger $\lambda \ne 0$.

2. Si $\lambda < 0$. Podemos llamar $\lambda = -k^2$, de forma que resolvemos la ecuacion
$$u''(x) = k^2 u(x) $$
La solucion de esta ecuacion son exponenciales $\mathrm{e}^{\pm kx}$.
Es decir

$$u(x) = A \mathrm{e}^{kx} + B \mathrm{e}^{-kx} $$
Debemos determinar $A,B$ de las condiciones de frontera. Es decir como $u(0)=u(\pi)=0$.

\begin{eqnarray}
A \mathrm{e}^0 + B \mathrm{e}^{0} = 0 \implies A = -B \\
\end{eqnarray}
Ahora para $x=\pi$

$$ -B \mathrm{e}^{k \pi} + B \mathrm{e}^{-k \pi} = -B ( \mathrm{e}^{k \pi} - \mathrm{e}^{-k \pi} ) = 0  $$
Tarea: Muestre que $ \mathrm{e}^{k \pi} - \mathrm{e}^{-k \pi} )$ solo es 0 para $k \ne 0$, de resto no puede ser 0. O sea que

$$ B = 0 $$
y por lo tanto para $k<0$ la solucion es $u(x)=0$. No nos interesa.

3. $\lambda > 0$: Podemos pensar que $\lambda = k^2$, queremos resolver
$L(u) = k^2 u$, $-u''(x) = k^2 u$, cuya solucion es

$$u(x) = A \cos kx + B \sin kx $$
Aplicamos las condiciones de frontera

$$u(0) = A \cos 0 + B \sin 0 = A = 0 $$
$$u(0) = 0 \implies  A = 0 $$
O sea que tenemos
$$u(x) = B \sin kx $$
$$u(x) = B_n \sin nx $$
donde $\lambda=n^2$.

Decimos que el problema
$$L(u_n) = \lambda u_n $$


donde $u_n = B_n \sin nx $ es la $n$ autofuncion. En la practica se normalizan las funciones. Es decir queremos que

$$\| B_n \sin nx \| = 1 $$

2. Que es la norma de una funcion?

### Numeros complejos.
* Numeros naturales $\mathbb{N}$ y el cero  $0,1,2, \cdots $
* Numeros enteros negativos $\mathbb{Z}^{-}$: Solucion de $x + n=0$, $n \in \mathbb{N}$.
* Numeros enteros $\mathbb{N} \cup \{ 0 \} \cup \mathbb{Z}^{-}$
* Numeros racionales $\mathbb{Q}$. Solucion de $ax=b$, $a,b \in \mathbb{Z}, a \ne 0$.
* Numeros irracionales $\mathbb{I}$. Solucion de $x^2 -a = 0$, $a \in \mathbb{N}$.
Aca no estan todos, falta $\pi$, $\mathrm{e}$, etc.

* Los reales $\mathbb{R}=\mathbb{Q} \cup \mathbb{I}$.
* Los imaginarios
$$x^2 + 1 = 0 $$
La solucion es

$$x = \sqrt{-1} = \mathrm{i} $$

* Los complejos :
Sean $a,b \in \mathbb{R}$ reales. Entonces $z= a + \mathrm{i} b$ es un numero complejo.

$$\mathbb{C} = \{ z | z = a + \mathrm{i} b \quad , \quad a,b \in \mathbb{R} \} $$

Los complejos se pueden graficar. Grafico a mano alzada.

El **conjugado** de $z=a + \mathrm{i}b$ es $\overline{z} = a - \mathrm{i} b$.
