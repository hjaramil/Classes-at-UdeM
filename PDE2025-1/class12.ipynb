La fuente la llamamos $g(x,t)$.  Definamos el operador

$$L(u) = \left [a \frac{\partial^2}{\partial t^2} + 2b \frac{\partial^2}{\partial x \partial x} + c \frac{\partial^2}{\partial x^2} + d \frac{\partial }{\partial t} + e \frac{\partial }{\partial x} + f \right ]u = g(x,t) $$

Si $g=0$ (no fuente) y no hay condiciones iniciales y de frontera, entonces
las soluciones $Lu=0$ son infinitas (si existen).
Por ejemplo si $u$ es solucion de $Lu=0$, entonces $\forall c \in \mathbb{R}$

$cu$ tambien lo es pues

$$L(cu) = c L(u) = 0 $$
Tambien, si $u_1, u_2$ son soluciones tambien lo es $u=u_1 + u_2$

$$L(u_1 + u_2) = L(u_1) + L(u_1) = 0  $$
Lo que hace que una solucion pueda ser unica es
* La fuente $g \ne 0$
* Condiciones iniciales y de frontera.

De alguna forma vamos a integrar doble sobre la fuente.

Piense  en la siguiente ecuacion simple

$$\frac{\partial^2 u}{\partial x \partial y} = g(x,y).  $$
Es claro que para poder integrar $u$ necesitamos resolver una integral doble sobre $x, y$.

**Ejemplo 3.3.1**: Considere el siguiente IVP:

$$
\begin{cases}
u_{tt} - c^2 u_{xx} = F(x,t) \quad , \quad -\infty < x < \infty \quad , \quad  t>0 \\
u(x,0) = \varphi(x) \\
\left .  \frac{\partial u(x,t)}{\partial t} \right |_{t=0}= \psi(x).
\end{cases}
$$

Para resolver este problema vamos a usar el
[Teorema de Green](https://en.wikipedia.org/wiki/Green%27s_theorem).
Sea $P=P(x,t)$, $Q=Q(x,t)$, funciones definidas en region  plana $\Omega$, con frontera $\Gamma$. Entonces

$$\iint_{\Omega} [Q_x(x,t) - P_t(x,t)] dx dt = \oint_{\Gamma} P(x,t) dx + Q(x,t) dt$$


Vamos a integrar la ecuacion de onda en el dominio de dependencia.


image.png

$$-\iint_{\Delta} F(x,t) dx dt = \iint_{\Delta} (c^2 u_{xx} - u_{tt}) dx dt$$

Le aplicamos a esto el teorema de Green con $Q=c^2 u_x$, $P=u_t$

$$-\iint_{\Delta} F(x,t) dx dt = \oint_{\Gamma} (u_t dx - c^2 u_x) dt=
\int_{B} + \int_R +\int L $$

1. Bottom (base del triangulo)
En la base $dt=0$.

$$\int_B u_t dx + c^2 u_x dt = \int_{x_0-t_0}^{x_0+t_0} u_t(x,0) dx = \int_{x_0 - t_0}^{x_0 + t_0} \psi(s) ds \tag{1} $$

2. Derecha (R): Lado derecho
$x + ct = \text{const}$. $dx = c dt$ o $dt=-dx/c$.
Voy cambiar el $dx$ por $dt$ y el $dt$ por dx$. Es decir

$$\int_R u_t (-c dt) + c^2 u_x (-dx/c) = -c \int_{R} u_t dt + u_x dx
= -c \int_R du = - c [u(x_0, t_0) - u(x_0+c t_0, 0)]  $$

Es decir

$$\int_R u_t dx + c^2 u_x dt = -c [ u(x_0, t_0) - \varphi(x_0 + c t_0)]  \tag{2} $$

3. A la izquierda (L):
Como $x - ct = \text{constante}$ entonces $dx = c dt$, o $dt=dx/c$ y reemplazando $dt$ en terminos de $dx$ y vice-versa tenemos

$$u_t dx + c^2 u_x dt = u_t (c dt) + c^2 u_x (dx/c) = c [u_t dt + u_x dx]$$

$$\int_L u_t dx + c^2 u_x dt = c \int_L u_t dt + u_x dx = c \int_L du $$

Es decir
$$\int_L u_t dx + c^2 u_x dt = c [u(x_0 -c t_0, 0) - u(x_0, t_0) $$

$$\int_L u_t dx + c^2 u_x dt = c [\varphi(x_0 -c t_0) - u(x_0, t_0) \tag{3}$$

La integral sobre la region es la suma (1) + (2) + (3)

$$-\iint_{\Delta} F(x,t) dx dt = \int_{x_0 - c t_0}^{x_0 + c t_0} \varphi(s) ds
 +  c[\varphi(x_0-ct_0) + \varphi(x_0 + c t_0) - 2u(x_0, t_0)$$
 Despesjamos de aca $u(x_0, t_0)$

 $$ u(x_0, t_0) = \frac{\varphi(x_0 + ct_0) + \varphi(x_0 - c t_0)}{2} + \frac{1}{2c} \int_{x_0 - c t_0}^{x_0 + c t_0} \psi(x) dx + \frac{1}{2c} \iint_{\Delta} F(x,t) dx dt.  $$


En general, para cualquier punto arbitrario $(x,y)$ escribimos

 $$ u(x, t) = \frac{\varphi(x + ct) + \varphi(x - c t)}{2} + \frac{1}{2c} \int_{x - c t}^{x + c t} \psi(x) dx + \frac{1}{2c} \iint_{\Delta} F(x,t) dx dt.  $$

Este solo un ejemplo de ecuacion de onda con fuente. Al final del curso vamos a estudiar un metodo general para resolver otras ecuaciones con fuente. Este metodo se le denomina **Green's Functions**.

## Cuerda semi-infinita.
Hasta el momento no hay fronteras $-\infty < x < \infty$.
Ahora vamos a considerar el problema
$$ u_{tt} = c^2 u_{xx} \quad , \quad x>0 \quad , t>0 $$
Al $x$ no ser negativo es semi-infinta $x \in [0, \infty)$
En $x=0$ hay condicion de borde. Ya el problema es un IBVP: Initial Boundary Value Problem

Se necesitan condiciones iniciales y de frontera

$$u(x,0) = \varphi(x)  \quad , \quad \frac{\partial u}{\partial t}(x,0)= \psi(x) \quad , \quad x > 0 .  $$
Falta la condicion de bordo $x=0$

$$u(x=0,t) = 0 , t > 0  $$
A esta condicion se le llama Direchlet.
Que pasa con la condicion inicial en la primera derivada ?

La solucion que tenemos es


$$u(x,t) = \frac{\varphi(x + ct) + \varphi(x - ct)}{2} + \frac{1}{2c} \int_{x-ct}^{x+ct} \psi(s) ds $$

Como la onda solo se propaga par $x>0$. Tenemos que la rama en $x - ct$ se cancela por que puede llegar a ser negativa.  $x-ct<0$.

La solucion general a la ecuacion de onda con velocidad constante es

$$u(x,t) = f(x-ct) + g(x+ct)  $$

Ahora aplicamos la condicion de frontera

$x=0$ tenemos
$$ $$
$$u(0,t) = f(-ct) + g(ct)= 0  \quad , t>0  $$

De forma que $f(-ct) = -g(ct)$
$f(-s) = -g(s) \tag{5a}$

Ahora si las condiciones iniciales

$$u(x,0) = f(x) + g(x) = \varphi(x) \tag{4}  $$
$$\frac{\partial }{\partial t} u(x,0) = -cf'(x) + c g'(x) =\psi(x) $$


Integremos la ecuacion anterior con respecto a $x$.

$$-c f(x) + c g(x) = \int_0^x \psi(s) ds  \tag{5}$$

Podemos eliminar $f$ de las ecuaciones (4) y (5)
A la (5) le sumamos la (4) escalada por $c$.

$$ 2 c g(x) =  c \varphi(x) + \int_0^x \psi(s) ds $$
Despejamos $g$

$$g(x) = \frac{\varphi(x)}{2} + \frac{1}{2c} \int_0^x \psi(s) ds $$

Recuerde que $f(-s) = -g(s)$ (de (5a)

Tenemos que para $x - ct < 0$


$$f(x- ct) = - g(ct - x) = -\frac{\varphi(ct-x)}{2} - \frac{1}{2c} \int_0^{ct -x} \psi(s) ds = -\frac{\varphi(ct -x)}{2} + \int_{ct -x}^0 \psi(s) ds$$

Entonces tenemos que para $0 < x < ct$

$$u(x,t) = f(x-ct) + g(x+ct) = \frac{\varphi(x+ct) - \varphi(ct-x)}{2} + \frac{1}{2c} \int_{ct-x}^{x+ct} \psi(s) ds $$

Podemos escribir esto de la siguiente forma

$$
u(x,t) =
\begin{cases}
\frac{\varphi(x+ ct) + \varphi(x-ct)}{2} + \frac{1}{2c} \int_{x - ct}^{x+ct} \psi(s) ds \quad , \quad  x > ct \\
\frac{\varphi(x + ct) - \varphi(ct - x)}{2} + \frac{1}{2c} \int_{ct-x}^{x+ct} \psi(s) ds \quad, \quad 0 < x < ct.
\end{cases}
$$
