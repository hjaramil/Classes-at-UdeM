## Funcion de Green para la ecuacion de onda.

Debemos resolver

$$u_{tt} - c^2 u_{xx} = F(x,t) $$
con condiciones iniciales homogeneas

$$u(x,0) = 0 \quad , \quad u_t(x,0) = 0  $$
y de frontera

$$u(\pm \infty, t) = 0. $$

La funcion de Green se calcula de la ecuacion

$$G_{tt} - c^2 G_{xx} = \delta(x-\xi) \delta(t - \tau) $$
$G= G(x, t; \xi, \tau) $.

Los pasos son
1. Transformada de Fourier de $x$ a $k$.

$$ \hat{G}_{tt} + c^2 k^2 \hat{C} dt = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} \delta(t-\tau). \tag{4} $$

2. Resolvemos la ecuacion homogenea. Es decir $t \ne \tau$.

$$\hat{G}_{tt} + c^2 k^2 \hat{G} = 0 . $$
Esta ecuacion tiene dos soluciones y a la hemos hecho antes.

$$\hat{G}(k, t; \xi, \tau) = A(k, \xi, \tau) \cos(ck(t - \tau)) +
B(k, \xi, \tau) \sin(ck(t - \tau)). \tag{5} $$

Dejo como tarea explicar por que el cambio de $t$ a $t-\tau$ no afecta el resultado de la ecuacion.

3. Imponemos las condiciones iniciales.
De la misma forma que hicimos en la ecuacion de calor, integramos sobre un intervalo pequenho $(\tau-\epsilon, \tau+ \epsilon)$.
Integramos la Ecuacion (4).



$$ \int_{\tau-\epsilon}^{\tau+\epsilon} \hat{G}_{tt} dt + \int_{\tau-\epsilon}^{\tau+ \epsilon} c^2 k^2 \hat{G} dt = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} $$

La segunda integral es 0 en el limite $\lim_{\epsilon \to 0}$ por que $\hat{G}$ es continua.

La primea integral y el ultimo termino producen

$$\hat{G}_t(k , \tau^+) - \hat{G}(k, \tau^-)  = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}.$$
La expresion del medio es 0 por causalidad. Es decir $\hat{G}(k, \tau^-)=0$.
Tenemos encones que

$$ \hat{G}_t(k , t=\tau) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}.$$
Ahora bien de la Ecuacion (5) con $t = \tau$

En $t = \tau$ ocurre que
$$\hat{G}(k, t=\tau; \xi)=0 $$

Reemplazando $t=\tau$ en (5)

$$\hat{G}(k, t=\tau; \xi, ) = A(k, \xi, \tau)   = 0  $$
O sea que $A=0$, y  de la Ecuacion (5)

$$ B(k, \xi, \tau) = B(k, \xi , \tau) c k \cos(0) $$.

## Ecuacion de onda en 2D.
Asumimos que $x \in [0, L_x]$, $y \in [0, L_y]$.
Entonces

\begin{eqnarray}
u_{tt} - c^2 ( u_{xx} + u_{yy}) &=& 0  \quad , \quad \text{Ecuacion de onda} \\
u(0,y, t) = u(L_x, y , t) &=& u(x,0,t) = u(x, L_y, 0 ) = 0  \quad , \quad \text{De frontera Dirichlet Homogeneas} \\
u(x, y, 0 ) &=& \varphi(x,y) \quad , \quad u_t(x,y,0) = \psi(x,y) = 0
\quad , \quad \text{condiciones iniciales.}
\end{eqnarray}

La idea de separacion de varibles es escribir la solucion como un triple producto

$$u(x,y,t) = X(x) Y(y) T(t). $$
Insertamos esto en la Ecuacion de ondaa y obtenemos

$$X(x) Y(y) T''(t) = c^2 ( X''(x) Y(y) T(t) + X(x) Y''(y) T(t). $$

Dividimos todo por $u(x,t)=X(x) Y(y) T(t)$

$$\frac{T''(t)}{c^2 T(t)} = \frac{X''(x)}{X(x)} + \frac{Y''(y)}{Y(y)}. $$
donde
$$ \frac{X''(x)}{X(x)} = \lambda^2 c^2 \tag{5b}$$ y
$$ \frac{Y''(y)}{X(y)} = \mu^2 c^2 \tag{5c} $$

El termino de la izquierda es solo funcion de $t$ y los de la derecha funcion de $x, y$. Entonces estos terminos solo cumplen con eso si son constantes.

$$\frac{T''}{T} = -c^2 (\lambda + \mu) = -\omega^2 \tag{6} $$
con $\lambda, \mu > 0$. Puedo renombrar $\lambda = k_x^2$, $\mu = k_y^2$.

Entonces escribimos la relacion de dispersion
de
$$-c^2 (\lambda + \mu) = -\omega^2 $$

$$k_x^2 + k_y^2 = \omega^2 . \tag{6a} $$
A esto se le llama la relacion de dispersion.

Encontramos entonces las siguientes tres ecuaciones ordinarias
de la (6).

$$T''(t) + \omega^2  T(t) = 0 \tag{7} $$

$$ X''(x) = \lambda c^2 X(x) = 0 $$
De (5b) y (5c)

$$X''(x) + \lambda c^2 X(x) = 0    \quad \text{y} \quad Y''(y) + \mu c^2 Y(y) = 0 . $$

Recordemos las condiciones de frontera
$$X(0) = X(L_x) = 0 \quad , \quad Y(0) = Y(L_y) = 0. $$

La solucion para $X$ y $Y$ son funciones seno. Revisando las notas anteriores.

Encontramos que

$$X(x) = C_1 \sin \left ( \frac{m \pi x}{L_x}  \right )  \quad , \quad Y(y) = C_2 \sin \bigg ( \frac{n \pi x}{L_x} \bigg ) $$

Podemos renombrar las autofunciones para mostrar que las constantes estan amarradas a $m,n$, son infinito

$$X_n(x) = C_{xn} \sin \left ( \frac{m \pi x}{L_x}  \right )  \quad , \quad
Y_n(y) = C_{yn} \sin \bigg ( \frac{n \pi x}{L_x} \bigg ) $$

Los productos de estas son

$$X_m(x) Y_n(y) = C_{xm} C_{yn} \sin \bigg ( \frac{m \pi x}{L_x}  \bigg )
\sin \bigg (  \frac{n  \pi y}{L_y} \bigg ). $$

Vamos ahora a resolver la Ecuacion  (7).

$$ T''(t) + \omega^2 T(t) = 0.$$
La solucion es de la forma

$$T(t) = A \cos (\omega t) + B \sin (\omega t). $$
como $u_t(x,y,0)=0.$

$$T'(t) =  - \omega A \sin (\omega t) + \omega B \cos (\omega t)
\bigg |_{t=0}  = B=0. $$

$$ T_n(t) = A_n \cos( \omega_n t). $$

La relacion de dispersion (6a) produce

$$\omega_{mn} = \sqrt{ \left (  \frac{m \pi}{L_x} \right )^2 + \left ( \frac{n \pi y}{L_y}^2  \right ) }$$

Usamos superposicion
de estas ondas

$$X_m(x) Y_n(x) T(t) = a_{mn} \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left (  \frac{n \pi y}{L_y}   \right ) \cos( w_{mn} t)  $$

para obtener

$$ u(x,y,t) = \sum_{m=1}^\infty \sum_{m=1}^{\infty} X_m(x) Y_n(x) T(t) =
\sum_{m=1}^\infty \sum_{n=1}^\infty a_{mn} \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left (  \frac{n \pi y}{L_y}   \right ) \cos( w_{mn} t)
\tag{8}.  $$


Para hallar $a_{mn}$ usamos la ortogonalidad de las autofunciones. Veamos
Llamemos

$$ \varphi_{mn} =  \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left (  \frac{n \pi y}{L_y}   \right ) \tag{8a} $$

\begin{eqnarray}
\langle \varphi_{mn}(x,y) , \varphi_{pq}(x,y)  \rangle &=&
\bigg \langle   \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left (  \frac{n \pi y}{L_y}   \right ),  \sin \left ( \frac{p \pi x}{L_x} \right ) \sin \left (  \frac{q \pi y}{L_y}   \right )  \bigg \rangle \\
&=& \int_0^{L_x} \int_0^{L_y}
  \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left (  \frac{n \pi y}{L_y}   \right )  \sin \left ( \frac{p \pi x}{L_x} \right ) \sin \left (  \frac{q \pi y}{L_y}   \right )  dx dy \\
  &=&  \int_0^{L_x} \sin \left ( \frac{m \pi x}{L_x}  \right ) \sin \left ( \frac{ p \pi x}{L_x} \right ) dx \; \int_0^{L_y} \sin \left ( \frac{n \pi y}{L_y}  \right ) \sin \left ( \frac{ p \pi y}{L_y} \right ) dy   \\
  &=& \frac{L_x}{2} \delta_{mp} \frac{L_y}{2} \delta_{nq}.
\end{eqnarray}

Usamos la condicion inicial en la Ecuacion (8)


$$ \varphi(x,y) = u(x,y,0) = \sum_{m=1}^\infty \sum_{m=1}^{\infty} X_m(x) Y_n(x) T(0) =
\sum_{m=1}^\infty \sum_{n=1}^\infty a_{mn} \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left (  \frac{n \pi y}{L_y}   \right )
 $$

 Es decir

 $$ \varphi(x,y) =  
\sum_{m=1}^\infty \sum_{n=1}^\infty a_{mn} \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left (  \frac{n \pi y}{L_y}   \right )  \tag{9}.
 $$
 Para hallar  los $a_{mn}$ proyectamos la condicion inicial sobre las bases con autofunciones $\varphi_{pq}$.  de la (8a).

$$\langle \varphi(x,y),   \varphi_{pq}(x,y) \rangle =  
\sum_{m=1}^\infty \sum_{n=1}^{\infty} a_{mn} \langle \psi_{mn}(x,y),  \psi_{pq}(x,y) \rangle  = \sum_{m=1}^\infty \sum_{n=1}^\infty a_{mn} \frac{L_x L_y}{4} \delta_{mp} \delta_{nq} =  a_{pq} \frac{L_x L_y}{4}.
$$
De donde (cambiamos $p$ por $m$ y $q$ por $n$

$$a_{mn} = \frac{4}{L_x L_y} \int_0^{L_x} \int_0^{L_y} \varphi(x,y)
\sin \left ( \frac{m \pi y}{L_x} \right ) \sin \left ( \frac{m \pi y}{L_y} \right )  dy dx.  $$

De forma que la solucion es:


$$ u(x,y,t) = \sum_{m=1}^\infty \sum_{n=1}^\infty a_{mn} \sin \left ( \frac{m \pi x}{L_x} \right ) \sin \left (  \frac{n \pi y}{L_y}   \right ) \cos( w_{mn} t) .  $$

Donde
$$a_{mn} = \frac{4}{L_x L_y} \int_0^{L_x} \int_0^{L_y} \varphi(x,y)
\sin \left ( \frac{m \pi y}{L_x} \right ) \sin \left ( \frac{m \pi y}{L_y} \right )  dy dx.  $$


Programemos esta solucin en `Python`




````python
# ecuacion de onda en 2D
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

# constantes
Lx, Ly = 1.0, 1.0
c=1.0
M,N=15,15

num_snapshot = 6
t_final = 2.0
frames=100 # para la animacion
dx = dy = 0.002

# create spatial grid
x = np.arange(0, Lx, dx)
y = np.arange(0, Ly, dy)
X, Y = np.meshgrid(x,y)

# condicion inicial Gaussiana
def phi(x,y):
    return np.exp(-400*((x-0.5)**2 + (y-0.5)**2))

# coeficientes de Fourier a_{mn}
def a_mn(m,n):
    A = 0
    for i in range(len(x)):
        for j in range(len(y)):
            A += phi(x[i],y[j])*np.sin(m*np.pi*x[i]/Lx)*np.sin(n*np.pi*y[j]/Ly)*dx*dy
    A *= 4/(Lx*Ly)
    return A

# suma de la solucion sobre los terminos m, n
def u(x,y,t):
    U = np.zeros_like(x)
    for m in range(1,M+1):
        for n in range(1,N+1):
            a_mn_value = a_mn(m,n)
            omega_mn = np.sqrt(m**2*np.pi**2/Lx**2 + n**2*np.pi**2/Ly**2)
            U += a_mn_value*np.sin(m*np.pi*x/Lx)*np.sin(n*np.pi*y/Ly)*np.cos(omega_mn*t)
    return U

# crear figura
fig, axes = plt.subplots(2, 3, figsize=(12,8))

# snapshots
times = np.linspace(0, t_final, num_snapshot)

for ax, t in zip(axes.ravel(), times):
    Z = u(X, Y, t)
    ax.contourf(X, Y, Z, levels=300, cmap='coolwarm')
    ax.set_title(f't = {t:.2f}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout
plt.show()


````

````python
from matplotlib.animation import FuncAnimation
# create animation
fix, ax = plt.subplots()

contour = ax.contourf(X, Y, u(X, Y, 0), levels=300, cmap='coolwarm')
ax.set_xlabel('x')
ax.set_ylabel('y')

def animate(t):
    ax.clear()
    Z = u(X, Y, t)
    contour = ax.contourf(X, Y, Z, levels=300, cmap='coolwarm')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    return contour

ani = FuncAnimation(fig, animate, frames=np.linspace(0, t_final, frames), interval=50,  blit=False)

# save the animation
ani.save('wave_equation.gif', writer='pillow')
plt.show()

````
