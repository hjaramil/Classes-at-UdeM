# Metodo de separacion de variables
Si quisieramos usar el metodo de d'Alembert y estamos por alla lejos, como en $1000 \ell$, necesitariamos como 100 ramas y se hace inmanejable y obvio es imposible escribir la solucion en el intervalo $(\infty, \infty)$ usando ramas de d'Alembert (necesitaria un numero infinito de ramas). Entonces el metodo de separacion de variables viene a salvarnos de este problema.


La **receta** es la siguiente

1. Asuma $u(x,t)=F(x) G(t)$. Es decir que $u(x,t)$ es el producto de dos funciones, una en espacio y otra en tiempo.

2. Inserte esta solucion en la ecuacion original (PDE). En este caso la ecuacion de onda. Esto le produce 2 ODE (Ordinary Differential Equations) ecuaciones ordinarias. Que puede resolver mas facilmente.

3. Resuelva las dos ODEs del paso 2 y aplicamos las condiciones de bordo (frontera). Con esto se producen un conjunto infinito de auto-funciones. (son bases del espacio)

4. Use el principio de superposicion para hallar una solucion mas general.
Esto es una suma al infinito (una serie) con coeficientes indeterminados.

5. Con las condiciones iniciales se halla el valor de los coeficientes indeterminados en el paso 4 y listo.

Veamos un ejemplo

**Ejemplo 4.2.1**: Considere el problema

$$
\begin{cases}
u_{tt} - c^2 u_{xx} = 0  \\
u(0,t) = u(\ell, t) = 0 \quad , \quad \text{condiciones de borde} \\
u(x,0) = \varphi(x) \quad, \quad u_t(x,0) = \psi(x) = 0 \quad, \quad \text{condiciones iniciales}
\end{cases}
$$

Usar el metodo de separacion de variables
1. Asumimos
$$u(x,t) = F(x) G(t)  $$

2.
$$u_{tt} = F(x) G''(t)  \quad , \quad u_{xx} = F''(x) G(t) $$
entonces, de la ecuacion de onda

$$F(x) G''(t) = c^2 F''(x) G(t) $$
De aca que

$$ \frac{F''(x)}{F(x)} = \frac{1}{c^2} \frac{G''(t)}{G(t)} = k $$
$k$ tiene que ser constante por que es imposible que dos funciones sean a la vez funciones de $x$ y de $t$ a no ser que sea una constante.

De forma

$$F''(x) = k F(x) \tag{3} $$
$$ G''(t) = k c^2 G(t). \tag{4} $$
Estas son las 2 ODE del paso 2.

3. Resolvamos la ecuacion (3) $F''(x) = k F(x)$.

Hay varias posibilidades

* Para el paso 3 opcion una
    *
$k=0$:
Entonces la ecuacion (3)
$$F''(x) = 0 \implies F(x) = C_1 x + C_2$$
Aplicamos las condiciones de  bordo
$u(x,t)= F(x) G(t)$

$u(0, t) = u(\ell, t) = 0$
De $u(0,t) = 0$

$$F(0) G(t) = 0   $$
$$ C_2 G(t) = 0   $$
como $G(t) \ne 0$, O sea que $C_2=0$
y
$$F(x) = C_1 x $$
de la condicion $u(\ell, t)=0$
$$F(\ell) \ G(t) = 0$$
$$C_1 \ell G(t) = 0$$
y como  $G(t) \ne 0 $ entonces
$$C_1 = 0$$
encontramos que $F(x) = 0$.
o sea que $k=0$ no nos sirve.

* Para el paso 3 opcion 2
    * \begin{eqnarray} k>0 \end{eqnarray}
Podemos, convenientemente decir $k = \lambda^2$.
Entonces la es ecuaciones (3) y (4) quedan

$$ F'' - \lambda^2 F = 0  \quad , \quad G'' - c^2 k^2 G = 0. $$
Resolvamos la primera. Sea $D=d/dx$, entonce tenemos

$$D^2 - \lambda^2 = (D - \lambda) (D + \lambda) $$

$$(D - \lambda) F = 0 \implies \frac{dF}{dx} = \lambda F \implies F= C_1 \mathrm{e}^{\lambda x} $$
$$(D + \lambda) F = 0 \implies \frac{dF}{dx} = \lambda F \implies F= C_2 \mathrm{e}^{-\lambda x} $$

Juntando las dos

$$ F(x) = C_1 \mathrm{e}^{\lambda x} + C_2 \mathrm{e}^{-\lambda x}$$

Ahora de las condiciones de frontera

$$F(0) = C_1 + C_2 = 0 \tag{4} $$
$$F(\ell) = C_1 \mathrm{e}^{\lambda \ell} + C_2 \mathrm{e}^{-\lambda \ell} = 0 \tag{5} $$

De la 4 tenemos que $C_2 = -C_1$. Insertamos eso en la (5) y vemos que

$$C_1 ( \mathrm{e}^{\lambda \ell} - \mathrm{e}^{-\lambda \ell}) = 0 $$
De tarea les deje que probaran que esto es 0 si y solo si
$\lambda=0$,  este caso ya esta descartado y ademas que
$\mathrm{e}^{\lambda \ell} - \mathrm{e}^{-\lambda \ell} >0$ para $\lambda \ne 0$. o sea
$C_1=0$
Entonces $C_1 = C_2 = 0$ y $F=0$ lo que no nos intersa. Entonces

* Caso 3 opcion 3
    * $$ k<0 < 0 $$
    Podemos asumir que $k=-\lambda^2>0$
    Las ecuaciones en F y G quedan

    $$F'' + \lambda F = 0  \quad , \quad G'' + c^2 \lambda^2 G = 0 $$

    Resolvemos la primera. Factorizamos el polinomio caracteristico. Donde $D=d/dx$.

    $$D^2 + \lambda^2= (D - \mathrm{i} \lambda)(D + \mathrm{i} \lambda)=0 $$
    Nos da os ecuaciones


$$(D - \lambda \mathrm{i}) F = 0  \implies \frac{dF}{dx} = \mathrm{i} \lambda F  \implies F = C_1 \mathrm{e}^{\mathrm{i} \lambda x} $$
$$(D + \lambda \mathrm{i}) F = 0  \implies \frac{dF}{dx} = -\mathrm{i} \lambda F  \implies F = C_2 \mathrm{e}^{-\mathrm{i} \lambda x} $$

Hacemos superposicion.

$$F(x) = C_1 \mathrm{e}^{\mathrm{i} \lambda x} + C_2 \mathrm{e}^{-\mathrm{i} \lambda x} $$

Recuerde:
$$
\sin \theta = \frac{\mathrm{e}^{\mathrm{i} \theta}- \mathrm{e}^{- \mathrm{i} \theta}}{2 \mathrm{i}}  \quad , \quad  
\cos \theta = \frac{\mathrm{e}^{\mathrm{i} \theta} +  \mathrm{e}^{- \mathrm{i} \theta}}{2}  
$$

Aplicamos las condiciones de frontera.

$$F(0) = C_1 + C_2 = 0 \tag{6} $$
$$F(\ell) = C_1 \mathrm{e}^{\mathrm{i} \lambda \ell}+ C_2 \mathrm{e}^{-\mathrm{i} \lambda \ell} = 0 \tag{7} $$

De la (6) tenemos que $C_2 = -C_1$. y usando esto en la (7)

 $$C_1 \mathrm{e}^{\mathrm{i} \lambda \ell} -C_1 \mathrm{e}^{-\mathrm{i} \lambda \ell} = 2 \mathrm{i} C_1 \sin \lambda \ell = 0 $$
 O sea que como $C_1 \ne 0$ entonces

 $$\sin \lambda \ell = 0 $$
 Esta ecuacion trigonometrica tiene un numero infinito de soluciones.

 $$\lambda \ell = n \pi \implies \lambda = \lambda_n = \frac{n \pi}{\ell}  $$
 donde $n \in \mathbb{Z}$. Esto me genera un conjunto infinito de autovalores con autofunciones

 $$\lambda_n = \frac{n \pi }{\ell }  \quad , \quad F_n(x) = 2C_1 \sin \frac{n \pi   x }{\ell}$$

 En este momento resolvemos la ecuacion para $G$. Recuerde la ecuacion

 $$D^2 + \lambda^2 c^2 = 0 \implies (D-\lambda c \mathrm{i})(D + \lambda c \mathrm{i}) = 0 $$

$$(D - \lambda c \mathrm{i}) G = 0  \implies \frac{dG}{dx} = \mathrm{i} \lambda c G  \implies G = C_3 \mathrm{e}^{\mathrm{i} c \lambda x} \tag{8} $$
$$(D + \lambda c \mathrm{i}) G = 0  \implies \frac{dG}{dx} = -\mathrm{i} c \lambda G  \implies G = C_4 \mathrm{e}^{-\mathrm{i} c \lambda x} $$

Por superposicion (2 veces)

$$G(t) = C_3 \mathrm{e}^{\mathrm{i} c \lambda t} + C_4 \mathrm{e}^{-\mathrm{i} c \lambda  t} \tag{6b} $$

Volvmos a aplicar las condiciones iniciale spara hallar $C_3, C_4$.

$$G(x, 0) = C_3 + C_4 = \varphi(x) \tag{7} $$
$$G_t(x,0) = C_3 \mathrm{i} c -C_4 \mathrm{i} c = 0  \tag{8} $$
De la (8) como $\mathrm{i} c \ne 0$, $C_3 = C_4$ y esto lo insertamos en la (6b)

$$ G(t) = 2 C_3 \cos \lambda c t $$
Ya cuando resovimos por $F$ encontramos que

$$\lambda = \lambda_n = \frac{n \pi}{\ell} $$
de forma que

$$  \lambda_n = \frac{n \pi }{n} \quad, \quad G(t) = 2 C_3 \cos \frac{nc \pi t }{\ell}   $$
    
Encontramos que

$$u_n(x,t) = F_n(x) G_n(t) = 4 \mathrm{i} C_1 C_3 \sin \frac{n \pi x }{\ell}  \cos \frac{n \pi c t}{\ell} $$

Por conveniencia definimos

$$a_n = 4 \mathrm{i} C_1 C_3 $$
Note que para cada $n$ tanto $C_1$ como $C_3$ puede cambiar.

Tenemos entonces que

$$u_n(x,t) = F_n(x) G_n(t) = a_n \sin \frac{n \pi x }{\ell} \cos \frac{n \pi c t}{\ell} $$

    

Ahora hacemos la **ultima superposicion**. Todas estas son soluciones. La superposicion tambien lo es

$$u(x,t) = \sum_{i=1}^{\infty} a_i \sin \frac{i \pi x}{\ell } \cos \frac{i \pi c t}{\ell} \tag{8a} $$

Por que no sobre $\mathbb{Z}$ en vez de sobre $\mathbb{N}$ como se esta haciendo en clase.

Tenemos que hallar $a_i, i=1,2, \cdots, \infty$
Para esto usamos la condicion inicial

$$u(x, 0) = \varphi(x) = \sum_{i=1}^\infty a_i \sin \frac{i  \pi x}{\ell} \tag{9}$$

En palabras: Para hallar $a_i$ multiplicamos (producto interno) $u(x,0)$ por cada una de las autofunciones $\sin j \pi x/\ell$ esto elimina todos los coeficientes excepto el $a_j$.

En abstracto. Tenemos

$$ v = \sum_{i=1}^\infty v_i \phi_i $$
si los $\phi_i$ son ortonormales, es decir

$$\langle \phi_i , \phi_j \rangle = \delta_{ij} $$
entonces al multiplicar $\langle v, \phi_j \rangle $ obtenemos

$$ \langle v, \phi_j \rangle  = \sum_{i=1}^\infty v_i \langle \phi_i, \phi_j \rangle = \sum_{i=1}^{\infty} v_i \delta_{ij} = v_j $$

Veamos sea

$$f_n(x) = \sin \frac{ n \pi x }{\ell} $$
Veamos que

$$\langle f_n(x), f_m(x) \rangle = 0 $$ para $n \ne m$
Es decir $f_n$ es ortogonal a $f_m$ (esto ya sabe por que el SLO es autoadjunto)

$$\langle f_n, f_m \rangle = \int_0^{\ell} \sin \frac{n \pi x }{\ell} \sin \frac{m \pi x }{\ell} dx $$

Ahora bien, de trigonometrica

$$\sin A \sin B = \frac{\cos (A - B) - \cos ( A + B)}{2} $$

De forma que

$$\langle f_n, f_m \rangle = \int_0^\ell \frac{\cos  (n-m) \frac{\pi x}{\ell} - \cos  (n+m) \frac{\pi x}{\ell}}{2} dx = \left .\frac12 \frac{\ell}{\pi(n-m)} \sin(n-m) \frac{\pi x}{\ell} - \frac12 \frac{\ell}{\pi (n+m)} \sin( n+m) \frac{\pi x}{\ell} \right |_0^\ell = 0. $$

Para la magnitud

$$ \| f_n \|^2 = \langle f_n , f_n \rangle = \int_0^{\ell} \frac{1 - \cos 2 \frac{n \pi x }{\ell}}{2} dx = \left .\frac{\ell}{2} - \frac{\ell}{2 n \pi x}\sin (2 \pi n x/\ell) \right |_0^\ell = \frac{\ell}{2}  $$

Encontramos que

$$\langle f_i, f_j \rangle = \frac{\ell}{2} \delta_{ij} $$

Ahora ya encontramos los $a_i$.
Multiplicamos (9) por $\sin j \pi x /\ell$
Entonces


$$\left \langle \varphi(x), \sin \frac{j \pi x}{\ell} \right \rangle = \sum_{i=1}^{\infty} a_i \left \langle \sin \frac{i \pi x}{\ell}, \sin \frac{j \pi x}{\ell}    \right \rangle = \sum_{i=1}^{\infty} a_i \delta_{ij}  \frac{\ell}{2} =  a_j \frac{\ell}{2} $$
De forma

$$a_j = \frac{2}{\ell} \left \langle \varphi(x), \sin \frac{j \pi x}{\ell} \right \rangle $$

Ahora reemplazmos este $a_j$ en la ecuacion (8a)
y encontramos


$$u(x,t) = \sum_{i=1}^{\infty} a_i \frac{\sin  i \pi x}{\ell} \cos \frac{i \pi c t}{\ell}  $$
con
$$a_i = \frac{2}{\ell} \left \langle \varphi(x), \sin \frac{i \pi x}{\ell} \right \rangle= \frac{2}{\ell} \int_0^\ell  \varphi(x) \sin \frac{i \pi x}{\ell} $$
