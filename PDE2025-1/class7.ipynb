# Empalme con la clase anterior:
Las caracteristicas para un sistema no lineal son:

$$ \frac{d x(\sigma)}{d \sigma} = F_p \quad , \quad  \frac{d y(\sigma)}{d\sigma} =  F_q \quad , \quad \frac{d u(\sigma)}{d \sigma} = p F_p + q F_q,  \tag{1} $$
donde $F=F(x,y,p,q,u)$ con $p=u_x, q=u_y$. Existen dos nuevas "incognitas", $p,q$ necesitamos ampliar el sistema de ecuacioens a 5.

**Ejemplo 2.2.8**: Asuma la ecuacion PDE no lineal $u_x^2 + u_y^2 = u$, resuelvala.

Solucion, en terminos de $p,q$
$$ F(x,y,u,p,q) = p^2 + q^2 - u = 0.$$
De aca que $F_p=2p$, $F_q=2q$,  las caracteristicas  (1) se reducen a

$$ \frac{d x(\sigma)}{d \sigma } = 2 p   \quad, \quad
 \frac{d y(\sigma)}{d \sigma } = 2 q   \quad , \quad
 \frac{d u(\sigma)}{d \sigma } = 2 u   \tag{0}
$$

Aparentemente ya tenemos todos los elementos para resolver el problema. Pero
**Quien es $p$? Quien es $q$?**
En la proxima clase, construirmos **dos** ecuaciones mas para poder resolver por $p,q$. O sea que necesitariamos **5 ecuaciones** en las variables
$x,y,p,q,u$.




parametricemos:

$$p = u_x( x(\sigma), y(\sigma)), q=u_y(x(\sigma), y(\sigma)) $$
Usamos la regla de la cadena

$$ p_\sigma = u_{xx} x_\sigma + u_{xy} y_\sigma = u_{xx} F_p + u_{xy} F_q $$
$$ q_\sigma = u_{yx} x_\sigma + u_{yy} y_\sigma = u_{yx} F_p + u_{yy} F_q $$

En terminos de $p$, $q$, como $p=u_x , p_x=u_{xx}, p_y=u_{xy}$, y $q=u_y, q_x= u_{yx}, q_y=u_{yy}$. Asumimos continuidad en las segundas derivadas

$$ u_{xy} = u_{yx} $$

$$p_\sigma = p_x F_p + p_y F_q \tag{1} $$
$$q_\sigma = q_x F_p + q_y F_q \tag{2} $$

Somo $u_{xy}=u_{yx}$ entonces $p_y = q_x$.

Ahora entonces de $F=F(x,y,u,p,q)$, con la regla de la cadena

Con respecto a $x$
$$F_x + p F_u + p_x F_p + q_x F_q $$
Con respecto a $y$
$$F_y + q F_u + p_y F_p + q_y F_q$$

Reemplazando (1) y (2) en las ultimas dos ecuaciones (recuerde $p_y=q_x$, y $p_x=q_y$)

$$ F_x + p F_u + p_\sigma = 0 $$
$$ F_y + q F_u + q_\sigma = 0 $$

Entonces encontramos las caracteristicas (de estas dos ultimas).

$$ \frac{dp}{d \sigma } = -F_x - p F_u $$
$$ \frac{dq}{d \sigma } = -F_y - q F_u $$

Concluimos que tenemos las 5 ecuaciones caracteristicas en $x,y,u,p,q$.

$$\begin{cases}  
\frac{d x(\sigma)}{d \sigma} = F_p(x,y,u,p,q)\\
\frac{d y(\sigma)}{d \sigma} = F_q \\
\frac{d u(\sigma)}{d \sigma} = p P_p + q F_q \\
\frac{dp}{d \sigma } = -F_x - p F_u  \\
\frac{dq}{d \sigma } = -F_y - q F_u
\end{cases} $$

Volvamos al ejemplo 2.2.8.
Obvservamos que $F_x=F_y=0$, $F_u=-1$
Las tres primeras ecuaciones ya  estan en (0). Las ultimas dos son


$$ \frac{dp}{d \sigma} = p $$
$$ \frac{dq}{d \sigma} = q $$

$$ p = k_1 \mathrm{e}^{\sigma} \quad, \quad q = k_2 \mathrm{e}^{\sigma} $$

Volvamos a las ecuaciones (0)
$$\frac{dx(\sigma)}{d \sigma} = 2 k_1 \mathrm{e}^{\sigma}  \quad , \quad \frac{d y(\sigma)}{d \sigma}  = 2 k_2 \mathrm{e}^{\sigma} \quad , \quad \frac{du(\sigma)}{d \sigma} =2 u $$

Resolvamos estas tres ecuaciones

$$\text{for } x(\sigma) : dx = 2 k_1 \mathrm{e}^{\sigma} d \sigma \implies x = 2 k_1 \mathrm{e}^{\sigma} + k_3  $$

$$\text{for } y(\sigma) : y = 2 k_2 \mathrm{e}^{\sigma} + k_4 $$

$$\text{for } u(\sigma): \frac{du}{u} = 2 d\sigma \quad , \quad \ln |u| = 2 \sigma + k_5 \implies u = k_5 \mathrm{e}^{2 \sigma} $$

Los pasos son

* Se comienza con una PDE no lineal de primer orden
* Se convierte en 5 ODE (ordinarias de primer orden)
* Se resuelven las 5 ODE pero quedan 5 constantes, $k_1, k_2, k_3, k_4, k_5$.

Como se halla en valor de estas constantes?
Hasta el momento,cuando eran lineales asumiamos una curva inicial
$x(\tau), y(\tau)$.
Bueno y como resolvemos el problema para las nuevas variables (funciones) $p,q$.
Definimos lo que llaman

**Defincion 2.2.3.2.1** (Characterstic Strips: Cintas caracteristicas)
En cada punto de la curva $x = x(\tau), y=y(\tau)$, $u=u(\tau)$, tomemos el plano tangente a la curva

$$u - u_0 = p_0(x - x_0) + q_0(y -y_0) $$
En cada punto de la caracteristica tenemos un plano tangente, y todos estos planos tangentes se les denomina **characteristic strips** (cintas caracteristicas).

Sabemos que

$$F(x_0, y_0, u_0, p_0, u_0) = 0  \tag{3}$$

En adicion a los datos $x_0=x_0(\tau), y_0=y_0(\tau), u_0=u_0(\tau)$, entonces
en cada punto $(x_0, y_0, u_0)$ tenemos dos curvas correspondientes

$$p_0 = p_0(\tau) \quad , q_0 = q_0(\tau) $$
A lo largo de las curvas tenemos, usando la regla de la cadena

$$ \frac{d u_0}{d \tau} =p_0 \frac{d x_0}{d \tau} + q_0 \frac{y_0}{d \tau} \tag{4}$$
Ademas para resolver el sistema $F(x_0, y_0, u_0, p_0, q_0)=0$ necesitamos que el jacobiano

$$ J = \begin{vmatrix} \frac{dx_0}{d\tau} & \frac{d y_0}{d \tau} \\ \frac{dF_0}{d p} & \frac{d F_0}{d q}  \end{vmatrix}   = \frac{d x_0}{d \tau} \frac{d F_0}{d q} - \frac{d y_0}{d \tau} \frac{d F_0}{dp} \ne 0. \tag{5} $$
Por el teorema de la funcion implicita.

Las ecuaciones (3), (4), (5) se conocemos como **generalized trasnversally conditions** (condiciones transversas generalizadas).

Vamos a ver que en caso de ecuaciones quasi-lineales, el sistema Jacobiano de alla es consistente con este problema.

$$ F(x,y,u,p,q ) = a(x,y,u) p + b(x,y,u) q - c(x,y,u) = 0$$
Este es el sistema quasi-lineal (de las clases de atras)
Entonces

$$F_p = a(x,y,u) \quad , \quad F_q=b(x,y,u) $$
entonces $J$ se reduce a

$$ \frac{d x_0}{d \tau} b - \frac{dy}{d \tau} a $$
vemos que este Jacobiano coincide con el Jacobiano (clases atras) para el caso quasi-lineal.


El algoritmo se convierte en

* **Paso #1**: Dado un IVP problem  con datos iniciales $(x_0, y_0, u_0)$ parametrizada por $\tau$, asumiento $\sigma=0$,

$$\begin{cases} F(x,y,u,p,q ) = 0 \\
x_0=x(0, \tau) \quad , \quad y_0 = y(0, \tau) \quad , \quad u_0=u_0(0, \tau)\end{cases} $$

Necesitamos dos curvas adicionales $(p_0, q_0)$ parametrizadas tambien por $\tau$, tales que $p_0=p_0(\tau)$ , $q_0=q_0(\tau)$ y

$$ F(x_0,y_0,u_0, p_0, q_0) = 0 $$
$$ \frac{d u_0}{d \tau} = p_0 \frac{d x_0}{d \tau} + q_0 \frac{dy_0}{d \tau} $$
Para escogencia de $(p_0, q_0)$ existe una unica solucion. Nosotros no proveemos $p_0, q_0$ ("a dedo") sino que ellas deben resolver las ecuaciones de arriba, es decir (3), (4).

* **Paso #2**: Se resuelven todas las 5 ecuaciones con las 5 condiciones iniciales.

En resumen:

* Comenzamos con la PDE y tres condiciones iniciales.
* La PDE se convierte en 5 ODEs. El sistema requiere resolver para 5 constantes.
* Agregamos las dos condiciones adicionales para $p_0, q_0$ y obtenemos un sistema de 5 ecuaciones algebraicas con 5 incognitas. La solucion de este sistmea me debe producer $u=u(x,y)$, que es la solucion integral.


Regresemos a nuestro ejemplo (lo reescribo)

**Ejemplo 2.2.9 continuacion del ejemplo 2.2.8**
Resuelva el IVP problem

$$\begin{cases}  
F(x,y,u, p, q) = p^2 + q^2 - u = 0 \\
x_0(\tau) = \tau \quad , y_0(\tau) = 0 \quad u_0(\tau) = \tau
\end{cases} $$




**Solucion**:
Tenemos para usar las condiciones trasversas

$$\frac{du_0}{d \tau} = 1  \quad , \frac{d x_0}{d \tau} = 1 \quad , \quad \frac{d y_0}{d \tau} = 0 $$

Necestamos  de las ecuaciones (3) y (4)
$$p_0^2 + q_0^2 = \tau  \tag{6} $$
$$1 = p_0 (1) + q_0 (0) = p_0$$
De forma que $p_0=1$.
Despejando $q_0$ de (6)

$$ q_0 = \pm \sqrt{\tau - 1 }  $$
Ya tenemos las dos condiciones iniciales que faltaban.

$$p_0=1 \quad , \quad q_0=\sqrt{\tau -1 } $$

Estamos listos para resolver las 5 ecuaciones algebraicas para $k_1,k_2,k_3,k_4,k_5$.
Las ecuaciones son

$$ p = k_1 \mathrm{e}^{\sigma} \quad , \quad q = k_2 \mathrm{e}^{\sigma} \quad , \quad u = k_5 \mathrm{e}^{2 \sigma} \quad , x = 2 k_1 \mathrm{e}^{\sigma} + k_3 \quad , \quad y = 2 k_2 \mathrm{e}^\sigma + k_4$$


Veamos
Como $p_0=1$ entonces $k_1=1$

Como $q_0=\sqrt{\tau - 1} $ entonces $k_2=\sqrt{\tau - 1}$
de forma que

$$p = \mathrm{e}^{\sigma} \quad , \quad q = \sqrt{ \tau -1} \mathrm{e}^\sigma $$

Buscamos las soluciones para $x,y,u$

$$x_0= 2 k_1 + k_3 $$
$$y_0= 2 k_2 + k_4 $$

Es decir
$$\tau = 2 + k_3  $$
$$0  = 2 \sqrt{\tau - 1} + k_4 $$

De aca que

$$ k_3 = \tau - 2 \quad , \quad k_4 =-2 \sqrt{\tau - 1}  $$

Como tenemos $k_3, k_4$ resolvemos por $x,y$

$$x = 2 \mathrm{e}^{\sigma} + \tau - 2 $$
$$y = 2 \sqrt{\tau -1} \mathrm{e}^\sigma -2 \sqrt{\tau -1}= 2 \sqrt{\tau -1}(\mathrm{e}^{\sigma} - 1) $$

Necesitamos invertir por $\sigma=\sigma(x,y)$, $\tau=\tau(x,y)$. Encontramos $\tau$ de la primera ecuacion, y en la segunda

$$\tau = x - 2 \mathrm{e}^\sigma + 2 $$


$$ y = 2 \sqrt{x - 2 \mathrm{e}^{\sigma} + 1}( \mathrm{e}^{\sigma} -1) $$

Dejamos como tarea hallar la solucion $\sigma=\sigma(x,y)$ de esta ultima ecuacion. Esto se obtiene resolviendo un polinomio de grado (3). Pueden usar WolframAlpha, una vez sepa $\sigma=\sigma(x,y)$ encuentran que

$$u(x,y) = \tau(x,y) \mathrm{e}^{2 \sigma(x,y)}  $$
Esta es la solucion.
