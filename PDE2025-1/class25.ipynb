## Transformada de Fourier


$$\hat{F}(k)  = \frac{1}{2 \pi} \int_{-\infty}^\infty F(x) \mathrm{e}^{-\mathrm{i} {k x}}  dx \tag{7} . $$


con ( de la (6))
$$ F(x) = \int_{-\infty}^\infty  \hat{F}(k) \mathrm{e}^{\mathrm{i} kx} dk$$

Notas:

1. $\hat{F}(k)$ es la transformada de Fourier de $F(x)$.
2. Otras notaciones que se usan

$$\hat{F}(k) = \mathcal{F}[ F(x)]  \quad ,
\quad \text{transformada de Fourier directa} $$
$$F(x) = \mathcal{F}^{-1}[ \hat{F} (k) ] \quad , \quad \text{transformada de Fourier inversa} $$

3. Note que

$$\mathcal{F} [ \mathcal{F}^{-1} [\hat{F}(k)] = \hat{F}(k) $$
y
$$\mathcal{F}^{-1}[ \mathcal{F} [ F(x)] = F(x) $$

4. Interpretacion: La expresion $\hat{F}(k)$ reprenta el **espectro** la descomposicion de la funcion $F(x)$ en los distintos numeros de onda. (un numero de onda es una frecuencia espacial que estudiamos hoy).

El proceso de ir de $F(x)$ a $\hat{F}(k)$ se llama analisis (analysis). Por ejemplo la luz blanca en un prisma se descompone en todos los numeros de onda que son colores. El proceso contrario, ir de $\hat{F}(k)$ a $F(x)$ se llama sintesis. Es la suma de todas frecuencias. El ejemplo es el de un disco con todos los colores del arco iris rotando rapidamente, el efecto es ver color blanco que es la suma de todos los colores.

Ejemplos
* El oido humano (como funciona el oido humado)
* El timbre de voz o de un instrumento. Que es el "timbre"
* El sintetizador electronico, para interpretar musica (guitarra en un teclado).

Otra aplicacion es la solucion de ecuaciones diferenciales, ordinarias o parciales. Por que convirten una ecuacion diferencial en algebraica que es mas de resolver. Los pasos son: se toma la transformada de Fourier de la ecuacion diferencial, se convierte en algebraica, se resulve la algebraica en el dominio $k$ y se toma la inversa para regresar al dominio $x$. En la proxima clase veremos esto.

Tambien se usa la transformada de Fourier para  calcular funciones de Green como veremos en esta clase.

Entonces tenemos dos dominios

$$F(x) \quad , \quad \text{espacio} $$
$$ \hat{F}(k) \quad , \quad \text{numero de onda} $$

Si no estamos en espacio sino en tiempo tenemos entonces otra notacion

$$F(t)  \quad , \quad \text{tiempo} $$
$$\hat{F}(\omega)  \quad , \quad \text{frecuencia} $$

Si estamos en tiempo el "transform pair" es

$$\hat{F}(\omega) = \frac{1}{2 \pi} \int_{-\infty}^\infty F(t) \mathrm{e}^{\mathrm{i} \omega t} dt $$

$$F(t) = \int_{-\infty}^{\infty} \hat{F}(\omega) \mathrm{e}^{-\mathrm{i} \omega t} d \omega $$


Hay distintas convenciones. Los coeficientes como los tenemos son
$\frac{1}{ 2 \pi}$ para la directa y $1$ para la inversa.
Es comun ver los coeficientes al contrario, es decir $1$ para directa y $\frac{1}{2 \pi}$ para  inversa.
A veces se reparten los dos coeficiente como $\frac{1}{\sqrt{2 \pi}}$ y $\frac{1}{\sqrt{2 \pi}}$. Otra version de coeficientes es que ambos sean 1 (cuando en vez de $\omega$ usas $2 \pi f$) Lo importante es que el producto de los dos coeficientes es simpre 1.

El signo del exponente del kernel (el kernel es $\mathrm{e}^{\mathrm{i} \omega t}$, en algunas versiones es "+" y en otras el "-".

Sin embargo si el signo en espacio es "+" en tiempo debe ser "-" para que halla causalidad. Causlidad es que si hay una explosion la onda se mueve hacia el infito. Anticausalidad es que la onda viene desde el infinito y colapsa en la fuente (implosion). Anticausalidad es como una pelicula tocando hacia atras.

Explicamos esto de causalidad con un ejemplo
La onda, en general satisface

$$u(x,t, k, \omega) = \mathrm{e}^{\mathrm{i} ( kx \pm \omega t)} $$

Tomemos el caso de una onda en los reales. Un coseno




$$u(x,t, k, \omega) = \cos(kx \pm \omega t) = \cos k \left ( x \pm \frac{\omega}{k} t  \right ) $$

Si nosotros insertamos esto en la ecuacion de onda encontramos

$$c^2 = \bigg (\frac{\omega}{k} \bigg )^2 $$
$$c = \frac{\omega}{k}$$.

Tambien podemos factorizar el $\omega$

$$u(x,t, k, \omega ) = \cos(kx \pm \omega t) = \cos \omega \bigg ( t \pm \frac{k}{\omega} x \bigg ) = \cos \omega \bigg( t  \pm \frac{x}{c} \bigg ) $$

Podemos ver los numeros de onda y frecuencias (circulares como)

$$\omega = 2 \pi f = \frac{2 \pi}{T} \quad , \quad \text{frecuencia circular}$$
$$k = \frac{2 \pi}{\lambda} = \frac{2 \pi}{c T} \quad , \quad \text{numero de onda} $$
aca $\lambda$ se llama **longitud de onda**.

## Transformada coseno integral:
El problema $Lu=-u''$ con $u(0)=0$ , $u'(0)=0$, nos lleva a al siguiente par de transformadas coseno en el dominio $[0, \infty)$

$$\hat{F}(k) = \int_0^\infty F(x) \cos k x dx $$
$$F(x) = \frac{2}{\pi} \int_0^{\infty} \hat{F}(k) \cos kx dk $$

Igualmente con senos, en el intervalo $[-L, L]$ en este caso resolvemos

$$F''(x) + \lambda^2 F = 0 $$
con $F(\pm L)=0$. se obtiene el par de transormadas

\begin{eqnarray}
F(x) = \int_{-\infty}^{\infty} F(x) \sin k x dx  \\
\hat{F}(k) = \frac{1}{2 \pi} \int_{-L}^{L} F(x) \sin k x dx  \\
\end{eqnarray}

Asi mismo hay una correspondencia entre ecuaciones diferenciales ordinarias con condiciones de bordo y transformadas. Algunas de ella son (nos las estudiamos aca)

* Tranformada de Hilbert.
* Tranformada de Laplace.
* Tranformada de Mellin.
etc.

Cuando las sumas son finitas la transformada es la tranformada $Z$ que es la DFT: Discrete Fourier Transform.

El ejemplo de Python es con cosenos. Esto me garantiza que funcion para funciones reales.

```python
import numpy as np
import matplotlib.pyplot as plt

# parametros para la senhnal
Fs=500  # frecuencia de mustreo
T = 1/Fs # periodo
L = 1000  # longitu de la senhal
t = np.arange(0, L)*T  # vector de tiempos

# definamos las comonengtes de la senhal. Frecuencia, fase, amplitud, analisis
frequencies = [5, 10, 20 ]
amplitudes = [1, 0.5, 0.2]
phases = [0, np.pi/4, np.pi/2]

# generamos las componentes de la suma individuales a_n exp(-k_n x)
individual_waves = []
summed_signal = np.zeros_like(t) # sintesis

# ciclo sobre cada componente
for freq, amp, phase in zip(frequencies, amplitudes, phases):
    wave = amp * np.cos(2*np.pi*freq*t + phase)
    individual_waves.append(wave)
    summed_signal += wave

# graficamos las componentes y su sintesis
plt.figure(figsize = (12, 12))
for i, wave in enumerate(individual_waves):
    plt.subplot(6, 1, i + 1)
    plt.plot(t, wave)
    plt.title(f'Componente {i + 1}')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')

# plot the sintesis
plt.subplot(6, 1, 4)
plt.plot(t, summed_signal)
plt.title('Senhal sintesis')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

plt.tight_layout()
```

```python
# dibujamos las senales de entrada en el dominio de la frecuencia. 3 spikes

# plot spikes to indicate the frequency components
plt.figure(figsize = (12, 12))

plt.subplot(6, 1 ,5)
# amplitude
plt.stem(frequencies, amplitudes)
plt.title('Componentes de la senhal')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')

# phase
plt.subplot(6, 1, 6)
plt.stem(frequencies, phases)
plt.title('Componentes de la senhal')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (rad)')

plt.tight_layout()
#
```

## La transformada de un "shift" (desplazamiento)
Sea $g(x) = f(x-s)$ sonde $s \in \mathbb{R}$. Calculemos la transformada de Fourier de $g$.

$$\mathcal{F}[ g(x) ] = \frac{1}{2 \pi} \int_{-\infty}^\infty g(x) \mathrm{e}^{-\mathrm{i k x} } dx = \frac{1}{2 \pi} \int_{-\infty}^\infty f(x-s) \mathrm{e}^{-\mathrm{i} k x} dx $$
Hacemos la sustitucion $u=x-s$, $du=dx$, $x=s+u$,

$$\mathcal{F}[ f(x-s)] = \frac{1}{2 \pi} \int_{-\infty}^\infty f(u) \mathrm{e}^{-\mathrm{i} k ( s  + u) } du = \mathrm{e}^{-\mathrm{i} k s} \frac{1}{2 \pi} \int_{-\infty}^\infty f(u) \mathrm{e}^{-\mathrm{i} k u} du = \mathrm{e}^{-\mathrm{i} k s} \mathcal{F}[ f(x)] $$

Es decir
$$\mathcal{F}[ f(x-s)] = \mathrm{e}^{-\mathrm{i} k s} \mathcal{F}[ f(x)].$$

## La transformada de Fourier de un Delta de Dirac

$$\mathcal{F}[\delta(x)] = \hat{\delta}(k) = \frac{1}{2 \pi} \int_{-\infty}^\infty \delta(x) \mathrm{e}^{-\mathrm{i} k x} dx = \frac{1}{2 \pi} $$

Es decir

$$\mathcal{F}[ \delta(x)] = \frac{1}{2 \pi} $$

Podemos combinar las dos propiedades anteriores en una asi

$$\mathcal{F}[ \delta(x - \xi)] = \mathrm{e}^{-\mathrm{i} k \xi} \mathcal{F}[\delta(x)] = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}  $$

El siguiente cuaderno ilustra el hecho de que el Delta de Dirac es la suma de todas las frecuencias, todas con fase 0.



```python
import numpy as np
import matplotlib.pyplot as plt

def sum_of_cosines_symetric(n, amplitude=1, time_interval=(-10,10), sampling_rate=1000):
    t = np.linspace(time_interval[0], time_interval[1], int(sampling_rate*(time_interval[1]-time_interval[0])), endpoint=False)

    # inicializacion de la senal
    signal = np.zeros_like(t)

    # sintesis: suma de cosenos con funciones simetrics
    for w in range(1, n+1):
        signal += amplitude*np.cos(2 * np.pi*w*t/10) # frecuencias positivas
        signal += amplitude*np.cos(2 * np.pi*(-w)*t/10) # frecuencias negativas

    # graph the resulting signal
    plt.figure(figsize=(10,5))
    plt.plot(t, signal, label=f'Sum of {2*n} cosines')
    plt.title(f'Sum of {n} cosines')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()

a=7
sum_of_cosines_symetric(n=80, amplitude=1, time_interval=(-a,a), sampling_rate=2000)
```
