# Empalme con la clase anterior: Funciones de Green (Green Functions).


Las funciones de Green son herramientas matematicas usadas para resolver ecuaciones diferenciales con fuente. La funcion de Green representa la solucion a una ODE o PDE (ordinaria o parcial) con fuente puntual (Dirac Delta). Es decir es la respueta impulso del sistema.


## Motivacion:
Vamos a comparar la solucion de sistemas lineales algebraicos con los sitemas lineales diferenciales.

Es decir, piense en el sistema

$$Ax = b \tag{2} $$
donde $A \in \mathbb{R}^{n \times n}$ y $x,y$ son vectores en $\mathbb{R}^n$.
Asumimos ademas que $A^{-1}$ existe.

Si $A^{-1}$ existe, podemos resolver la Ecuacion (2)  para cualquier "fuente" $b$.  Por ejemplo si conocemos $A^{-1}$ y la "fuente" es $y$ entonces la
ecuacion $A x = y$ se resuelve como

$$x = A^{-1} y. $$

De otro lado tenemos operadores de la forma


$$L u = f \tag{3}  $$
donde $L$ es un operador lineal (diferencial, una ODE o una PDE)
quremos resolver esta ecuacion. Encontrar $u$.

Vamos a retornar a la ecuacion (2) con distintas "fuentes"


$$A x_1 = \mathrm{e}_1 $$
$$A x_2 = \mathrm{e}_2 $$
$$ \vdots $$
$$A x_n = \mathrm{e}_n \tag{4} $$
donde $\mathrm{e}_i$ es un vector canonico con 0 en todas partes menos en la componente $i$. O sea $(\mathrm{e}_i)_j = \delta_{ij}$ .

Construyamos ahora la ecuacion

$$A [x_1 | x_2 | \cdots | x_n ] = [ \mathrm{e}_1 | \mathrm{e}_2 | \cdots | \mathrm{e}_n]  = I $$

De forma que tenemos el sistema

$$AB = I $$
donde
$$ B= [x_1 | x_2 | \cdots| x_n ]$$
Claramente $B=A^{-1}$.

En notacion indizal las ecuaciones (4) las podemos escribir como

$$A( x_i)_j = \delta_{ij} \quad i=1,2 , \cdots , n \quad , \quad j = 1,2, \cdots, n  \tag{5}  $$

Vamos a extender esta idea al operador (4). Queremos hallar algo como

$$L G(x, \xi) = \delta(x - \xi) \tag{6}$$

Comparamos las ecuaciones (5) con (6).

* La (5) es en el discreto finito, la (6) en el continuo infinito.
* $$ A ↔  L $$
* $$ G ↔  A^{-1} $$
* $$ i ↔  x $$
* $$ j ↔  \xi $$
* $$ \delta_{ij} ↔  \delta(x- \xi) $$



Queremos resolver la Ecuacion (6).

Recordemos como se resuelve la Ecuacion (2)

$$Ax = b  $$
Multiplicamos a ambos lados por $A^{-1}$, a izquierda

$$A^{-1}Ax = A^{-1} b  $$
$$x = A^{-1}b  $$

Que pasa si tenemos el operador inverso $L^{-1}(x, \xi)=G(x, \xi)$ (que es la funcion de Green).
Al sistema

$$Lu = f $$
lo multiplicamos (producto interno) a ambos lados por $L^{-1}=G$, es decir


$$u(x) = \langle G(x, \xi), f(\xi) \rangle = \int_{-\infty}^{\infty} G(x, \xi) f(\xi) d \xi. \tag{7} $$

Veamos este resultado de otra forma.

Recuerde la superposicion en el delta de Dirac

$$f(x) = \int_{-\infty}^\infty f(\xi) \delta(x - \xi) d \xi \tag{8} $$
y como

$$L G(x, \xi) = \delta(x-\xi) $$
insertamos esto en la Ecuacion (8)

$$f(x) = \int_{-\infty}^\infty f(\xi) L [G(x, \xi)] d \xi  $$
Como $L$ es lineal y $f$ no depende de $\xi$ la podemos mover adentro

$$f(x) = \int_{-\infty}^\infty  L [ f(\xi) G(x, \xi)] d \xi  $$
Sacamos el $L$ de la integral (por linealidad)
tenemos que como $Lu = f

$$L u(x) = L \left [  \int_{-\infty}^\infty f(\xi) G(x, \xi) d \xi \right ] $$

Ahora aplicamos $L^{-1}$ a ambos lados y obtenemos

$$ u(x) =   \int_{-\infty}^\infty f(\xi) G(x, \xi) d \xi \tag{9}$$

Vemos que la Ecuacion (9) coincide con la Ecuacion (7).

Esto funciona bien para ODE (Ecuaciones diferenciales ordinarias), pero que pasa con PDE? Por ejemplo una ecuacion

$$ Lu = f \tag{10} $$
$u=u(x,t)$.

Aca necesitamos dos deltas, uno para $x$ y uno para $t$ y la fuente **puntual** esta dada por


$$\delta(x - \xi) \delta(t - \tau) $$

Ya la funcion necesita 4 variables (si $u$ es funcion de $n$ variables, la funcion de Green necesita $2n$ variables).

Tenemos entonces que para resolver (10) tenemos

$$L G(x, t; \xi, \tau) = \delta(x - \xi) \delta(t - \tau). $$

De la misma forma que resolvimos $Lu=f$, $u=u(x)$ ahora resolvemos
$Lu=f$ pero con $u=u(x,t)$. Es decir que si $G(x, t; \xi, \tau)=L^{-1}$ entonces


$$u(x,t) = \langle G(x,t, . , . ), f(x,t) \rangle = \int_{-\infty}^{\infty} \int_{-\infty}^\infty G(x,t; \xi, \tau) f(\xi, \tau) d \xi d \tau. \tag{9} $$


Se deja como tarea al estudiante obtener esta ecuacion.
Existe algo llamado **causalidad**. Esto dice que mas alla del frente de onda, no hay nada. La funcion de Green mas alla del frente de onda $t > \tau$ es 0.

Entonce la intergral (9) se puede escribir como


$$u(x,t) = \langle G(x,t, . , . ), f(x,t) \rangle = \int_{-\infty}^{\infty} \int_{-\infty}^t G(x,t; \xi, \tau) f(\xi, \tau) d \xi d \tau.  $$

Vemos la importancia de tener las funciones de Green $G$. Si las tenemos podemos resolver el problema $Lu=f$, en cualquier numero de variables. Aunque en esta clase mostramos solo el caso de una $u=u(x)$ y dos $u=u(x,t)$ variables.

La pregunta en este momento es, como encontrar las funciones de Green?

Existen varios metodos para encontrar las funciones de Green

* Para dominios finitos $[a,b]$ o funciones periodicas los metodos de **Fourier** son los mas eficientes. Tambien para metodos en dominio infinito $(-\infty, \infty)$.

* Para dominios con condiciones de frontera, separacion de variables y la transformada de Laplace es conveniente. La transformada de Laplace tambien se usa para metodos semi-infinito $(-\infty, a)$ o $(a, \infty)$.

* El metodo de imagenes o ecauciones integrales puede ser util dependiendo de la geometria del problema y las condiciones de contorno. (donde hay simetrias).

El metodo que vamos a utilicar en esta clase es el metodo de la **Transformada de Fourier**.


## Transformada de Fourier
La transformada de Fourier es muy util para resolver ecuaciones diferenciales (ODE, PDE) por que converite estas ecuaciones en ecuaciones algebraicas, las cuales son mas faciles de resolver. Veamos la teoria de transformadas.


### Teoria de Transformadas.
Asumamos operadores en un espacio de funciones. Por ejemplo $S=L^2[a,b]$.
Que es el espacio de funciones que su cuadrado tienen integral de Lebesgue
en el dominio $[a,b]$, $a<b$, $a,b \in \mathbb{R}$.  Definimos el problema autovalor/vector como

$$L \varphi_n = \lambda_n \varphi_n $$
donde $L$ es un operadore lineal en $S$, $\varphi_n$ es una **autofuncion** y $\lambda_n$ es un autovalor.

Asumamos ademas que el espacio de autofunciontes $\varphi_n$ es completo, es decir, es una base para $S$ y ortogonal. Si es ortogonal, lo podemos normalizar para que sea ortonormal. De forma que tenemos que para cualquier funcion $f$
real en el dominio $[a,b]$

$$
\begin{cases}
f(x) = \sum_{i=1}^\infty c_i \varphi_i(x) \\
c_i = \langle f, \varphi_i \rangle = \int_D f(x) \varphi_i(x) dx
\end{cases}
\tag{11}
$$
A la expresion (11) lo llamamos un **transfor pair** (una par transormado)
y los coeficientes $c_i$ se llaman **Fourier Coefficientes**.

Observe que la expresion (11) tiene la siguiente propiedad.

1. Si tenemos los coeficientes $c_i$ podemos reconstruir $f$.
2. Si tnemos $f$ podemos construir los coeficientes $c_i$.

**Ejemplo 5.2.1**: Recuerden el ejemplo 4.2.1 donde resolvimos la ecuacion de ondaa en una dimension en el intervalo $[0 \; \ell]$
Obtuvimos el siguiente sistema

$$
\begin{cases}
F'' + \lambda^2 F = 0 \\
F(0) = F(\lambda) = 0.
\end{cases}
$$
Cuando resolvimos este sistema encontramos un numero infinito de soluciones de la forma

$$\lambda_n = \frac{n \pi}{\ell} $$
$$F_n = \sqrt{ \frac{2}{\ell}} \sin \frac{n \pi x}{\ell}. $$

El conjunto $\{ F_n \}_{n=1}^{\infty}$ es ortonormal. Es decir

$$ \langle F_i, F_j \rangle = \delta_{ij} $$
ademas es completo. Podemos escribir la condicional inicial como

$$\varphi(x) = \sum_{i=1}^{\infty} c_i F_i $$

$$c_i = \langle \varphi(x), F_i \rangle = \int_0^{\ell} \varphi(x) F_i(x) dx $$

Mas concretamente

$$\varphi(x) = \sum_{i=1}^\infty c_i \sqrt{\frac{2}{\ell}} \sin \frac{n \pi x}{\ell}  $$
$$c_i = \sqrt{\frac{2}{\ell}} \int_0^\ell \sin \frac{n \pi}{\ell} \varphi(x) dx. $$

Veamos otro ejemplo con condiciones de frontera periodicas en $[-L,L]$.

**Ejemplo 5.2.2**: Consideremos la ecuacion de onda en el intervalo $[-L,L]$.
Obtenemos, mediante separacion de variables.

$$
\begin{cases}
F''(x) + \lambda^2 F(x) = 0   \\
F(-L) = F(L) \quad, \quad F'(-L) = F'(L).
\end{cases}
$$
Dos soluciones de la ecuacion ordianaria son

$$\mathrm{e}^{\mathrm{i} \lambda x}  \quad , \quad \mathrm{e}^{-\mathrm{i} \lambda x} $$
Caulquier combinacion lineal de estas tambien es solucion.

$$F(x) = C_1 \mathrm{e}^{\mathrm{i} \lambda x} + C_2 \mathrm{e}^{-\mathrm{i} \lambda x} $$

Apliquemos las condiciones de frontera $F(-L)=F(L)$.

$$C_1 \mathrm{e}^{-\mathrm{i} \lambda L} + C_2 \mathrm{e}^{\mathrm{i} \lambda L} = C_1 \mathrm{e}^{\mathrm{i} \lambda L} + C_2 \mathrm{e}^{\mathrm{-\mathrm{i} \lambda L}} $$

Es decir



Es decir

$$-2 \mathrm{i} C_1 \sin \lambda L = -2 \mathrm{i} C_2 \sin \lambda L$$

$$ (C_2 - C_1) \sin \lambda L = 0  \tag{12}$$

Ahora apliquemos la segunda ecuacion de frontera $F'(L)=F'(-L)$.

$$\mathrm{i} \lambda C_1 \mathrm{e}^{-\mathrm{i} \lambda L} - \mathrm{i} \lambda C_2 \mathrm{e}^{\mathrm{i} \lambda L} =\mathrm{i} \lambda C_1 \mathrm{e}^{\mathrm{i} \lambda L} - \mathrm{i} \lambda \mathrm{C_2} \mathrm{e}^{-\mathrm{i} \lambda L} $$

Agrupamos los $C_1$ a la izquierda y los $C_2$ a la derecha

$$-2 \mathrm{i} \lambda C_1 \sin \lambda L = 2 \mathrm{i} \lambda C_2 \sin \lambda L$$
Es decir

$$\lambda(C_1 + C_2) \sin \lambda L=0  \tag{13}$$

Una forma de resolver el sitema $2 \times 2$ (12, 13) esto es $C_1=C_2=0$. Pero esto produce la solucion trivial $u(x,t)=0$. De forma que mas bien la solucion debe ser $\sin \lambda L = 0$, es decir  $\lambda L = n \pi$.
y tenemos un numero infinito de soluciones (con autovalor)

$$\lambda_n = \frac{n \pi}{L} $$

Entonces concluimos que

$$F(x) = C_1 \mathrm{e}^{\mathrm{i \frac{n \pi x}{L}}} + C_2 \mathrm{e}^{-\mathrm{i} \frac{n \pi x}{L}} $$

Si hacemos superposicion entre $-\infty, \infty$ tenemos

$$ F(x) = \sum_{n=-\infty}^{\infty} a_n \mathrm{e}^{-\mathrm{i} \frac{n \pi x}{L}}\tag{14} $$

Veamos que el conjunto de funciones $\{\mathrm{e}^{\mathrm{i} \frac{n \pi}{L}} \}$  es ortogonal. Asumamos $n \ne m$


\begin{eqnarray}
\langle \mathrm{e}^{\mathrm{i} \frac{n \pi x}{L}}, \mathrm{e}^{\mathrm{i} \frac{m \pi x}{L}} \rangle &=& \int_{-L}^L \mathrm{e}^{\mathrm{i} (n - m) \frac{\pi }{L} x} dx \\
&=& \left .  \frac{L}{\mathrm{i}(n-m) \pi} \mathrm{e}^{\mathrm{i}(n-m) \frac{\pi x}{L}} \right |_{-L}^L \\
&=&   \frac{L}{\mathrm{i}(n-m) \pi} [
     \mathrm{e}^{\mathrm{i}(n-m) \pi } +
     \mathrm{e}^{-\mathrm{i}(n-m) \pi} ] \\
&=&   \frac{2 \mathrm{i} L}{\mathrm{i}(n-m) \pi}  \sin (n-m) \pi \\
&=& 0
\end{eqnarray}

Si $n=m$

$$\langle  \mathrm{e}^{\mathrm{i} \frac{n \pi x}{L}},  \mathrm{e}^{\mathrm{i} \frac{n \pi x}{L}} \rangle  = \int_{-L}^L dx = 2L $$

O sea que

$$\langle  \mathrm{e}^{\mathrm{i} \frac{n \pi}{L}},  \mathrm{e}^{\mathrm{i} \frac{m \pi}{L}} \rangle  = 2 L \delta_{nm} $$

El factor de normalizacion es $\sqrt{2 L}$. Tenemos entonces que

Para extraer $a_n$ de (14) proyectamos, es decir
$$\langle F(x), \mathrm{e}^{\mathrm{i} \frac{k \pi}{L} x} \rangle =
\sum_{n=-\infty}^\infty a_n \langle  \mathrm{e}^{\mathrm{i} \frac{n \pi}{L}},  \mathrm{e}^{\mathrm{i} \frac{k \pi}{L}} \rangle = \sum_{n=-\infty}^\infty a_n 2 L \delta_{nk}= 2 a_k L, $$
y asi

$$a_n=  \frac{1}{2L} \int_{-L}^L F(x)  \mathrm{e}^{-\mathrm{i} \frac{n \pi}{L}} dx $$
