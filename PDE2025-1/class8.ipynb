# Algoritmo general para resolver una PDE no lineal de primer orden
* Queremos resolver una ecuacion no lineal en derivadas parciales de primer order. Se escribe como $F(x,y,u,p,q)=0$.

* Convertirmos la ecuacion a un sistema de 5 ecuaciones en derivadas ordinarias (ODE)

* Con las condiciones iniciales obtenemos un sistema de 5 ecuaciones algebraicas.
3 condiciones iniciales para $x,y,u$ y con ellas resolvemos un sistema de 2x2 para hallar $p_0,q_0$.

* Luego de resolver para $\sigma(x,y)$, $\tau(x,y)$ invertimos este sistema para que quede en $u=(x,y)$

**Ejemplo 2.2.10**(ecuacion eikonal: Gobierna los tiempos en la propagacion de ondas. Trazado de rayos): La formula por primera vez Willian Rowan Hamilton (1805-1865). Es importante en la solucion de tiempos de viajes en optica geometrica.  En particula en la sismica se usa para trazado de rayos.
Asumimos una propagacion de onda en un medio con velocidad $v=v(x,y)$.
Queremos medir el tiempo $T$ desde que se genero la explosion hasta cada punto del medio $(x,y)$.  El gradiente de $T$ esta dado por

$$\nabla T = \left [ \frac{\partial T}{\partial x}, \frac{\partial T}{\partial y}  \right ]  $$
A esto se conoce como el **slowness vector** (vector de lentitud). La ecuacion eikonal dice que la norma del **slowness vector** es el inverso de la velocidad.  Es decir


$$ \| \nabla T \| = \frac{1}{v^2(x,y)} $$
Si $p=\partial T/\partial x$, $q=\partial T/\partial y$ entonces esto dice que

$$p^2 + q^2 = \frac{1}{v^2(x,y)} $$

Vamos a asumir que la velocidad es constante (por facilismo). Entonces tenemos

$$p^2 + q^2 = \frac{1}{v^2} $$

$$ F(x,y,u,p,q)= p^2 + q^2 - \frac{1}{v^2} =0$$
En la clase anterior vimos el sistema de caracteristicas 5x 5

$$ \frac{dx}{\sigma}=F_p \quad , \quad  \frac{dy}{d\sigma}=F_q \quad , \quad \frac{du}{d \sigma} = p F_p + q F_q \quad , \quad \frac{dp}{d \sigma} = -F_x - p F_u \quad , \quad \frac{dq}{d\sigma} = -F_y - q F_u \tag{1}  $$

Como
$$F_p=2p \quad , \quad F_q=2 q \quad , \quad F_x=F_y=F_u=0 $$
entonces

$$p F_p + q F_q = p (2p) + q(2q) = 2( p^2 + q^2) = \frac{2}{v^2}$$

Ahora bien, de (1)


$$\frac{dx}{d \sigma } = 2 p \quad , \quad \frac{dy}{d \sigma} = 2 q \quad , \quad \frac{du}{d \sigma} = \frac{2}{v^2} \quad , \quad \frac{dp}{d \sigma} = \frac{dq}{d \sigma} = 0 . $$

Las condiciones iniciales las ponemos a medida que resolvemos el problema.

De las ultimas dos ecuaciones tenemos que

$$p= p_0 \quad, \quad q = q_0 $$ (constantes)

De las dos primeras ecuaciones

$$ x = x_0 + 2 p_0 \sigma \quad , \quad y=y_0 + 2 q_0 \sigma  $$
Asumiendo que conocemos las condiciones iniciales $p_0, q_0, x_0, y_0$
invertimos $\sigma, \tau$ en terminos de $x,y$.

Resolvamos por $u$,
$$ u = u_0 + \frac{2 \sigma}{v^2}$$

Veamos, tenemos 3 ecuaciones

$$p_0^2 + q_0^2 = \frac{1}{v^2}  $$
$$ x = x_0 + 2 p_0 \sigma $$
$$ y = y_0 + 2 q_0 \sigma $$
se resolver a mano facilmente:
1. Despejar $\sigma$ de las 2 ultimas
2. Reemplazar en la primera.
3. Soluciona por $p_0$ y $q_0$ y obtiene

\begin{eqnarray}
p_0 = \frac{x-x_0}{v \sqrt{ (x-x_0)^2 + (y - y_0)^2}} \\
q_0 = \frac{y-y_0}{v \sqrt{ (x-x_0)^2 + (y - y_0)^2}} \\
\sigma = \frac{v \sqrt{(x-x_0)^2 + (y - y_0)^2}}{2}
\end{eqnarray}

De forma que
$$u = u_0 + \frac{\sqrt{(x-x_0)^2 + (y-y_0)^2}}{v} \tag{2} $$

Para nuestro caso $u=T$, que es tiempo de viaje.
El tiempo de viaje entre la "fuente" $(x_0, y_0)$ a qualquier punto del medio (plano) es $espacio/velocidad + tiempo_{inicial}$, $d/v + T_0$. y como

$$d = \sqrt{(x-x_0)^2 + (y- y_0)^2} $$
entonces

$$T = T_0 + \frac{\sqrt{(x-x_0)^2 + (y - y_0^2)}}{v} $$
pero si $u=T$ vemos esta es la solucion (2).

Vamos a interpretar $p_0,q_0$. De la ecuaciones

$$x(\sigma) = x_0 + 2 p_0 \sigma \quad , \quad y(\sigma)=y_0 + 2 q_0 \sigma $$
Tenemos

$$\frac{y-y_0}{x-x_0}= \frac{q_0}{p_0} $$
Entonces $q_0/p_0$ es la pendiente de la recta (rayo) que une la fuente con el destino.
Mas interpretacines. A las curvas de nivel (donde $T=constante$) se les llama **wavefronts** (frente de onda) y los caminos que unen $(x_0, y_0)$ con el destino $(x,y)$ se les llama rayos.

## Ecuaciones PDE (Partial Differential Equations) de segundo orden (lineas).
Ejemplo: onda, calor, Poisson, Laplace.

La forma general de una ecuacion PDE de segundo orden (lineal o no)

$$F= F(x_1, x_2, \cdots, x_n,u, u{,_1},u_{,_2}, \cdots,u{,_n}, u_{,11}, u_{,12}, \cdots, u_{,nn}) = 0   $$

Donde
$$u_{,i}= \frac{\partial u}{\partial x_i} \quad i=1,2 \cdots, n $$
$$u_{,ij}= \frac{\partial^2 u}{\partial x_i \partial x_j} \quad i=1,2 \cdots, n \quad , \quad j = 1,2 \cdots, n$$

Tambien se usa la notacion $u_{xx}, u_{xy}$, etc para decir $\partial^2 u/\partial x \partial x$, o $\partial^2 u / \partial x \partial y$

Por ejemplo una ecuacion lineal PDE de segundo orden en dos variables se escribe

$$L[u] = a u_{xx} + 2b u_{xy} + c u_{yy} + d u_x + e u_y + fu - g $$
El 2 es por que si esta ecuacion fuera algebraica (como vamos a ver) seria una forma cuadratica con discriminante $b^2 - 4 ac$. Pero con ese dos el discriminante es $4 b^2 - 4 ac = 4(b^2 - ac)$, y como el 4 no influye en el signo, lo ignoramos y definimos el **discriminante**

Pensemos en la solucion de la ecuacion $a x^2 + 2 bx + c=0$

$$x = \frac{-2b \pm 2 \sqrt{b^2-ac}}{2a}  = \frac{b \pm \sqrt{b^2 - ac}}{a} = \frac{b \pm \sqrt{D}}{a}$$

como
$$D = b^2 - a c $$
Las ecuaciones tienen una clasificacion de acuerdo al signo del determinante.
Veamos

1. **Elipticas**: En este caso $D<0$. Un ejemplo es la ecuacion de Laplace

$$ u_{xx} + u_{yy} = 0. $$
Verifique esto con $a=1, b=0, c=1$

2. **Parabolicas**: Aca $D=0$. La clasica es la ecuacion de calor.

$$ u_t = \alpha u_{xx} $$
Verifique esto con $a=1, b=0, c=0$

3. **Hiperbolicas**: aca $D>0$, la clasica es la ecuacion de onda

$$ u_{tt} - \alpha u_{xx} = 0, \alpha > 0  $$
Verifique esto con $a=-\alpha, b=0, c=1$
De forma $$b^2 - ac = 0^2 -(-\alpha) 1 = \alpha > 0. $$
