# Solucion a la Ecuacion de Laplace en tres dimensiones.
Asumimos una funcion $u(x,y,z)$ que satisface la ecuacion de Laplace con condiciones de frontera en la caja $[0, L_x] \times [0, L_y] \times [0, L_z]$.
La ecuacion de Laplace es:

$$\nabla^2 u = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} + \frac{\partial^2 u}{\partial z^2} = 0.  \tag{1} $$

Todas las condiciones de frontera son 0, excepto por la condicion en $z=L_z$ donde
$$u(x,y,L_z) = \varphi(x,y). $$

Como siempre asumimos separacion de varibles. Es decir

$$u(x,y,z) = X(x) Y(y) Z(z). $$
Insertamos esto en  (1) y dividmos por $u(x,y,z)$ y obtenemos, como antes

$$\frac{X''(x)}{X(x)}  + \frac{Y''(y)}{Y(y)} + \frac{Z''(z)}{Z(z)} = 0. $$
Con el mismo argumento que hemos venido usando tenemos que

$$ \frac{X''(x)}{X(x)} = - \lambda^2  \quad , \quad \frac{Y''(y)}{Y(y)} = -\mu^2 $$
y

$$\frac{Z''(z)}{Z(z)} = \gamma^2 = \lambda^2 + \mu^2 . $$

Usamos las condiciones de frontera y encontramos, como antes
las siguientes autofunciones con sus respectivos autovalores.

$$\lambda_n = \frac{m \pi}{L_x} \quad , X_n(x) = C_1 \sin \frac{m \pi x}{L_x} $$
$$\mu_n = \frac{n \pi}{L_x} \quad , Y_n(x) = C_2 \sin \frac{n \pi y}{L_y} $$

Ahora vamos a hallar la solucion en $Z$. La Ecuacion es

$$Z''(z) n- \gamma^2 Z(z) = 0. $$
Sabemos que esta ecuacion tiene dos soluciones linealmente independientes que son $\sinh \gamma z , \cosh \gamma z$
La solucion general esta dada por

$$Z(z) = C_3 \sinh \gamma z + C_4 \cosh \gamma z. $$
Aplicamos las condiciones de frontera. Como $Z(0)=0$

$$Z(0) = C_4=0 $$
y
ademas

$$\gamma_{mn} = \sqrt{\left ( \frac{ m \pi}{L_x} \right )^2 + \left (
    \frac{n \pi}{L_y} \right )^2 } $$

$$Z(z) = C_3 \sinh \sqrt{\left ( \frac{ m \pi}{L_x} \right )^2 + \left (
    \frac{n \pi}{L_y} \right )^2 } z  $$

Sigue
* Superposicion
* Condicion de frontera en $Z$ para hallar los coeficientes indeterminados.

$$u(x,y,z) = \sum_{m=1}^\infty \sum_{n=1}^\infty  a_{mn} \sin \frac{m \pi x}{L_x} \sin \frac{n \pi y}{L_y}   \sinh \sqrt{\left ( \frac{ m \pi}{L_x} \right )^2 + \left (
    \frac{n \pi}{L_y} \right )^2 } z \tag{2} $$

Para hallar los coeficientes $a_{mn}$ proyectamos la condicion de frontera en $z$ sobre esta solucion.

En $z$ tenemos que en $z=L_z$, tenemos que $u(x,y,L_z) = \varphi(x,y)$.

$$
u(x,y,L_z) = \sum_{m=1}^\infty \sum_{n=1}^\infty  a_{mn} \sin \frac{m \pi x}{L_x} \sin \frac{n \pi y}{L_y}   \sinh \sqrt{\left ( \frac{ m \pi}{L_x} \right )^2 + \left (
    \frac{n \pi}{L_y} \right )^2 } L_z = \varphi(x,y)
$$

Definamos las autofunciones

$$\varphi_{mn} = \sin \frac{m \pi x}{L_x} \sin \frac{ n \pi y}{L_y} $$
sabemos

$$\langle \varphi_{mn}, \varphi_{pq} \rangle =  \delta_{mp} \delta_{nq} \frac{L_x L_y}{4}. $$

Con esto hagamos la proyeccion de $\varphi(x,y)$ a $\varphi_{pq}$.

\begin{eqnarray}
\langle \varphi(x,y) , \varphi_{pq} \rangle &=& \sum_{m=1}^\infty \sum_{n=1}^\infty a_{mn} \left \langle  \varphi_{mn}
 \sinh \sqrt{\left ( \frac{ m \pi}{L_x} \right )^2 + \left (
    \frac{n \pi}{L_y} \right )^2 } L_z, \varphi_{pq}
\right \rangle  \\
&=& \sum_{m=1}^\infty \sum_{n=1}^\infty a_{mn}  \sinh \sqrt{\left ( \frac{ m \pi}{L_x} \right )^2 + \left (
    \frac{n \pi}{L_y} \right )^2 }  L_z \delta_{mp} \delta_{nq} \frac{L_x L_y}{4} \\
    &=& a_{pq}  \sinh \sqrt{\left ( \frac{ p \pi}{L_x} \right )^2 + \left (
    \frac{q \pi}{L_y} \right )^2 }  \frac{L_x L_y}{4}
\end{eqnarray}

De donde, y cambiano $p \to m $, $q \to n$.

$$a_{mn} = \frac{4}{L_x L_y}  \sinh \sqrt{\left ( \frac{ m \pi}{L_x} \right )^2 + \left (
    \frac{n \pi}{L_y} \right )^2 }   \int_0^{L_x} \int_0^{L_y} \varphi(x,y)
    \sin \frac{m \pi x}{L_x} \sin \frac{n \pi y}{L_y} dx dy \tag{3} $$
La solucion es la Ecuacion (2), donde $a_{mn}$ esta determinado por la Ecuacion (3).
