**Ejemplo 5.2.2**: Consideremos la ecuacion de onda en el intervalo $[-L,L]$.
Obtenemos, mediante separacion de variables.

$$
\begin{cases}
F''(x) + \lambda^2 F(x) = 0   \\
F(-L) = F(L) \quad, \quad F'(-L) = F'(L).
\end{cases}
$$
Dos soluciones de la ecuacion ordianaria son

$$\mathrm{e}^{\mathrm{i} \lambda x}  \quad , \quad \mathrm{e}^{-\mathrm{i} \lambda x} $$
Caulquier combinacion lineal de estas tambien es solucion.

$$F(x) = C_1 \mathrm{e}^{\mathrm{i} \lambda x} + C_2 \mathrm{e}^{-\mathrm{i} \lambda x} $$

Apliquemos las condiciones de frontera $F(-L)=F(L)$.

$$C_1 \mathrm{e}^{-\mathrm{i} \lambda L} + C_2 \mathrm{e}^{\mathrm{i} \lambda L} = C_1 \mathrm{e}^{\mathrm{i} \lambda L} + C_2 \mathrm{e}^{\mathrm{-\mathrm{i} \lambda L}} $$

Es decir



Es decir

$$-2 \mathrm{i} C_1 \sin \lambda L = -2 \mathrm{i} C_2 \sin \lambda L$$

$$ (C_2 - C_1) \sin \lambda L = 0  \tag{2}$$

Ahora apliquemos la segunda ecuacion de frontera $F'(L)=F'(-L)$.

$$\mathrm{i} \lambda C_1 \mathrm{e}^{-\mathrm{i} \lambda L} - \mathrm{i} \lambda C_2 \mathrm{e}^{\mathrm{i} \lambda L} =\mathrm{i} \lambda C_1 \mathrm{e}^{\mathrm{i} \lambda L} - \mathrm{i} \lambda \mathrm{C_2} \mathrm{e}^{-\mathrm{i} \lambda L} $$

Agrupamos los $C_1$ a la izquierda y los $C_2$ a la derecha

$$-2 \mathrm{i} \lambda C_1 \sin \lambda L = 2 \mathrm{i} \lambda C_2 \sin \lambda L$$
Es decir

$$\lambda(C_1 + C_2) \sin \lambda L=0  \tag{3}$$

Una forma de resolver el sitema $2 \times 2$ (2, 3) esto es $C_1=C_2=0$. Pero esto produce la solucion trivial $u(x,t)=0$. De forma que mas bien la solucion debe ser $\sin \lambda L = 0$, es decir  $\lambda L = n \pi$.
y tenemos un numero infinito de soluciones (con autovalor)

$$\lambda_n = \frac{n \pi}{L} $$

Entonces concluimos que

$$F(x) = C_1 \mathrm{e}^{\mathrm{i \frac{n \pi}{L}}} + C_2 \mathrm{e}^{-\mathrm{i} \frac{n \pi}{L}} $$

Si hacemos superposicion entre $-\infty, \infty$ tenemos

$$ F(x) = \sum_{n=-\infty}^{\infty} a_n \mathrm{e}^{-\mathrm{i} \frac{n \pi}{L}}\tag{4} $$

Veamos que el conjunto de funciones $\{\mathrm{e}^{\mathrm{i} \frac{n \pi}{L}} \}$  es ortogonal. Asumamos $n \ne m$


\begin{eqnarray}
\langle \mathrm{e}^{\mathrm{i} \frac{n \pi}{L}}, \mathrm{e}^{\mathrm{i} \frac{m \pi}{L}} \rangle &=& \int_{-L}^L \mathrm{e}^{\mathrm{i} (n - m) \frac{\pi}{L} x} dx \\
&=& \left .  \frac{L}{\mathrm{i}(n-m) \pi} \mathrm{e}^{\mathrm{i}(n-m) \frac{\pi x}{L}} \right |_{-L}^L \\
&=&   \frac{L}{\mathrm{i}(n-m) \pi} [
     \mathrm{e}^{\mathrm{i}(n-m) \pi } +
     \mathrm{e}^{-\mathrm{i}(n-m) \pi} ] \\
&=&   \frac{2 \mathrm{i} L}{\mathrm{i}(n-m) \pi}  \sin (n-m) \pi \\
&=& 0
\end{eqnarray}

Si $n=m$

$$\langle  \mathrm{e}^{\mathrm{i} \frac{n \pi}{L}},  \mathrm{e}^{\mathrm{i} \frac{n \pi}{L}} \rangle  = \int_{-L}^L dx = 2L $$

O sea que


$$\langle  \mathrm{e}^{\mathrm{i} \frac{n \pi}{L}},  \mathrm{e}^{\mathrm{i} \frac{m \pi}{L}} \rangle  = 2 \delta_{nm} $$

El factor de normalizacion es $\sqrt{2 L}$. Tenemos entonces que

Para extraer $a_n$ de (4) proyectamos, es decir
$$\langle F(x), \mathrm{e}^{\mathrm{i} \frac{k \pi}{L} x} \rangle =
\sum_{n=-\infty}^\infty a_n \langle  \mathrm{e}^{\mathrm{i} \frac{n \pi}{L}},  \mathrm{e}^{\mathrm{i} \frac{k \pi}{L}} \rangle = \sum_{n=-\infty}^\infty a_n 2 L \delta_{nk}= 2 a_k L, $$
y asi

$$a_n=  \frac{1}{2L} \int_{-L}^L F(x)  \mathrm{e}^{-\mathrm{i} \frac{n \pi}{L}} dx \tag{5} $$

## Transformada de Fourier
Los autovalores $n \pi/L$ son **numeros de onda** (wave numbers). Son modos discretos en la suma. Los podemos llamar

$$k(n, L) = \frac{n \pi}{L} $$

Queremos el limite cuando $L \to \infty$, usando calculo diferencial tenemos que

$$\Delta k  = k(n+1) - k(n)= \frac{\pi}{L} $$
Tomando el limite cuando $L \to \infty$

$$dn \to \frac{L dk}{\pi} $$
Donde el $n$ es discreto y $k$ es continuo. Entonces cambiamos la suma (4) por una integral

$$\sum_{n=-\infty}^\infty  \to \int_{-\infty}^\infty \frac{L}{\pi} dk $$

Es decir

$$F(x) = \int_{-\infty}^\infty \frac{a_k L}{\pi} \mathrm{e}^{\mathrm{i} k x} dk \tag{6} $$

Redefinimos el coeficiente, de la Ecuacion (5)

$$\frac{a_k L}{\pi} = \frac{L}{\pi}\frac{1}{2 L} \int_{-L}^L F(x) \mathrm{e}^{-\mathrm{i} \frac{\pi x}{L}} = \frac{1}{2 \pi } \int_{-L}^L F(x) \mathrm{e}^{-\mathrm{i} k x} dx $$
Recuerden el limite cuando $L \to \infty$, y renombrando

$$a_k L/\pi = \hat{F}(k)$$.
y

y cuando $L \to \infty$

$$\hat{F}(k)  = \frac{1}{2 \pi} \int_{-\infty}^\infty F(x) \mathrm{e}^{-\mathrm{i} {k x}}  dx \tag{7} . $$


con ( de la (6))
$$ F(x) = \int_{-\infty}^\infty  \hat{F}(k) \mathrm{e}^{\mathrm{i} kx} dk$$

Notas:

1. $\hat{F}(k)$ es la transformada de Fourier de $F(x)$.
2. Otras notaciones que se usan

$$\hat{F}(k) = \mathcal{F}[ F(x)]  \quad ,
\quad \text{transformada de Fourier directa} $$
$$F(x) = \mathcal{F}^{-1}[ \hat{F} (k) ] \quad , \quad \text{transformada de Fourier inversa} $$

3. Note que

$$\mathcal{F} [ \mathcal{F}^{-1} [\hat{F}(k)] = \hat{F}(k) $$
y
$$\mathcal{F}^{-1}[ \mathcal{F} [ F(x)] = F(x) $$

4. Interpretacion: La expresion $\hat{F}(k)$ reprenta el **espectro** la descomposicion de la funcion $F(x)$ en los distintos numeros de onda. (un numero de onda es una frecuencia espacial que estudiamos hoy).

El proceso de ir de $F(x)$ a $\hat{F}(k)$ se llama analisis (analysis). Por ejemplo la luz blanca en un prisma se descompone en todos los numeros de onda que son colores. El proceso contrario, ir de $\hat{F}(k)$ a $F(x)$ se llama sintesis. Es la suma de todas frecuencias. El ejemplo es el de un disco con todos los colores del arco iris rotando rapidamente, el efecto es ver color blanco que es la suma de todos los colores.

Ejemplos
* El oido humano (como funciona el oido humado)
* El timbre de voz o de un instrumento. Que es el "timbre"
* El sintetizador electronico, para interpretar musica (guitarra en un teclado).

Otra aplicacion es la solucion de ecuaciones diferenciales, ordinarias o parciales. Por que convirten una ecuacion diferencial en algebraica que es mas de resolver. Los pasos son: se toma la transformada de Fourier de la ecuacion diferencial, se convierte en algebraica, se resulve la algebraica en el dominio $k$ y se toma la inversa para regresar al dominio $x$. En la proxima clase veremos esto.

Entonces tenemos dos dominios

$$F(x) \quad , \quad \text{espacio} $$
$$ \hat{F}(k) \quad , \quad \text{numero de onda} $$

Si no estamos en espacio sino en tiempo tenemos entonces otra notacion

$$F(t)  \quad , \quad \text{tiempo} $$
$$\hat{F}(\omega)  \quad , \quad \text{frecuencia} $$

Si estamos en tiempo el "transform pair" es

$$\hat{F}(\omega) = \frac{1}{2 \pi} \int_{-\infty}^\infty F(t) \mathrm{e}^{\mathrm{i} \omega t} dt $$

$$F(t) = \int_{-\infty}^{\infty} \hat{F}(\omega) \mathrm{e}^{-\mathrm{i} \omega t} d \omega $$


Hay distintas convenciones. Los coeficientes como los tenemos son
$\frac{1}{ 2 \pi}$ para la directa y $1$ para la inversa.
Es comun ver los coeficientes al contrario, es decir $1$ para directa y $\frac{1}{2 \pi}$ para  inversa.
A veces se reparten los dos coeficiente como $\frac{1}{\sqrt{2 \pi}}$ y $\frac{1}{\sqrt{2 \pi}}$. Lo importante es que el producto de los dos coeficientes es simpre 1.

El signo del exponente del kernel (el kernel es $\mathrm{e}^{\mathrm{i} \omega t}$, en algunas versiones es "+" y en otras el "-".

Sin embargo si el signo en espacio es "+" en tiempo debe ser "-" para que halla causalidad. Causlidad es que si hay una explosion la onda se mueve hacia el infito. Anticausalidad es que la onda viene desde el infinito y colapsa en la fuente (implosion). Anticausalidad es como una pelicula tocando hacia atras.

Explicamos esto de causalidad con un ejemplo
La onda, en general satisface

$$u(x,t, k, \omega) = \mathrm{e}^{\mathrm{i} ( kx \pm \omega t}) $$

Tomemos el caso de una onda en los reales. Un coseno




$$u(x,t, k, \omega) = \cos(kx \pm \omega t) = \cos k \left ( x \pm \frac{\omega}{k} t  \right ) $$

Si nosotros insertamos esto en la ecuacion de onda encontramos

$$c^2 = \bigg (\frac{\omega}{k} \bigg )^2 $$
$$c = \frac{\omega}{k}$$.

Tambien podemos factorizar el $\omega$

$$u(x,t, k, \omega ) = \cos(kx \pm \omega t) = \cos \omega \bigg ( t - \frac{k}{\omega} x \bigg ) = \cos \omega \bigg( t - \frac{x}{c} \bigg ) $$

Podemos ver los numeros de onda y frecuencias (circulares como)

$$\omega = 2 \pi f = \frac{2 \pi}{T} \quad , \quad \text{frecuencia circular}$$
$$k = \frac{2 \pi}{\lambda} = \frac{2 \pi}{c T} \quad , \quad \text{numero de onda} $$
aca $\lambda$ se llama **longitud de onda**.
