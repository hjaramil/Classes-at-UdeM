## Coeficientes variables.
Considere la ecuacion (**ojo esta vez en $x,t$)

$$Lu = a(x,t) u_{tt} + 2 b(x,t) u_{tx} + c(x,t) u_{xx} + d(x,t) u_t + e(x,t) u_x + f(x,t) u = g(x,t) \quad , \quad x \in \mathbb{R} \quad , t > 0 \tag{1a} $$

Por el momento ignoramos las condiciones iniciales. Asumimos
que en un punto $(x,t)=(x_0,t_0)$

$$D = \left .  b^2(x,t) - a(x,t) c(x,t) \right |_{x=x_0, t=t_0}  > 0 $$

### Cambio de coordenadas.
Queremos hallar unas nuevas coordenadas $\xi, \eta$ tales que en las nuevas coordenadas los terminos puros (en $a,c$) se cancelen (sean 0).

Si esto sucediera entonces para $u=u(x(\xi, \eta), y(\xi, \eta))$

$$Lu = 2 b^*(x,t) u_{tx}  + d^*(x,t) u_t + c^*(x,t) u_x + f^*(x,t) u = g^*(x,t) \quad , \quad x \in \mathbb{R} \quad , t > 0 $$


la transormacion en las nuevas variables es

$$\xi = \xi(x, t) $$
$$\eta = \eta(x, t) $$

Queremos el jacobiano no sea 0.

$$J = \begin{vmatrix} \xi_t & \xi_x \\ \eta_t & \eta_x  \end{vmatrix} = \xi_t \eta_x - \xi_x \eta_t \ne 0 $$

Vamos a usar la regla de la cadena.

* Primeras derivadas con respecto a $t, x$.

$$ u_t = u_\xi \xi_t + u_\eta \eta_t \tag{2}  $$
$$ u_x = u_\xi \xi_x + u_\eta \eta_x \tag{3}  $$

* Segunda derivada $tx$.

\begin{eqnarray}
u_{tx} &=& (u_t)_x = (u_\xi \xi_t + u_\eta \eta_t)_x \\
&=& (u_\xi \xi_t)_x + (u_\eta \eta_t)_x \\
&=& (u_\xi)_x \xi_t + u_\xi \xi_{tx} + (u_\eta)_x \eta_t + u_\eta \eta_{tx}
\end{eqnarray}
Es decir
$$ u_{tx} = (u_{\xi \xi} \xi_x + u_{\eta \xi} \eta_x ) \xi_t + u_\xi \xi_{tx} + (u_{\eta \xi} \xi_x + u_{\eta \eta} \eta_x) \eta_t + u_\eta \eta_{tx} \tag{4} $$

De la misma forma

$$ u_{tt} = (u_{\xi \xi} \xi_t + u_{\eta \xi} \eta_t ) \xi_t + u_\xi \xi_{tt} + (u_{\eta \xi} \xi_t + u_{\eta \eta} \eta_t) \eta_t + u_\eta \eta_{tt} \tag{5} $$

$$ u_{tt} = u_{\xi \xi} \xi_t^2 + 2 u_{\xi \eta} \xi_t \eta_t + u_{\eta \eta} \eta_t^2 + u_\xi \xi_{tt} + u_\eta \eta_{tt}  \tag{6} $$



y cambiando los indices $t$ por $x$

$$ u_{xx} = u_{\xi \xi} \xi_x^2 + 2 u_{\xi \eta} \xi_x \eta_x + u_{\eta \eta} \eta_x^2 + u_\xi \xi_{xx} + u_\eta \eta_{xx}  $$

Insertamos (2),(3),(4),(5),(6) en la (1a) y agrupamos terminos, obtenemos


$$a^* u_{\xi \xi} + 2 b^* u_{\eta \xi} + c^* u_{\eta \eta} + d^* u_\xi + e^* u_\eta + f^* u = g^* $$

Donde

\begin{eqnarray}
a^* &=& a \xi_t^2 + 2 b  \xi_t \xi_x + c \xi_x^2 \\
b^* &=& a \xi_t \eta_t + b(\xi_t \eta_x + \xi_x \eta_t ) + c \xi_x \eta_x \\
c^* &=& a \eta_t^2 + 2 b \eta_t \eta_x  + c \eta_x^2 \\
d^* &=& a \xi_{tt} + b \xi_{tx} + c \xi_{xx} + d \xi_t + e \xi_x \\
e^* &=& a \eta_{tt} + b \eta_{tx} + c \eta_{xx} + d \eta_t + e \eta_x \\
f^* &=& f \\
g^* &=& g \\
\end{eqnarray}

Vamos a mostrar enseguida que el discriminante en estas nuevas coordenadas es tambien > 0.  O sea que se converva la propiedade de hiperboloidal.

\require{cancel}
Para evaluar esto vamos a considerar el siguiente orden
$b^2, ac, a^2, c^2, ab, bc$.
\begin{eqnarray}
D^* &=& (b^*)^2 - (a^* c^*)  \\
&=& b^2[ ( \xi_t \eta_x) + \xi_x \eta_t)^2 - 4 \xi_t \xi_x \eta_x \eta_x]
-ac [\xi_t^2 \eta_x^2 + \xi_x^2 \xi_t^2 - 2 \xi_t \eta_t \xi_x \eta_x ]
+ \require{\cancel}\cancel{a^2[\xi_t^2 \eta_t^2 - \xi_t^2 \eta_t^2]} + \cancel{c^2[\xi_x^2 \eta_x^2 - \xi_x^2 \eta_x^2]} + \cancel{ab[ 2 \xi_t \eta_t ( \xi_t \eta_x + \xi_x \eta_t} -2 \xi_t^2 \eta_t \eta_x - 2 \eta_t^2 \xi_t \xi_x]
+ \cancel{bc[2 \xi_x \eta_x(\xi_t \eta_x + \xi_x \eta_t) - 2 \eta_x^2 \xi_t \xi_x - 2 \xi_x^2 \eta_t \eta_x]} \\
&=& b^2( \xi_t \eta_x - \xi_x \eta_t)^2 - ac(\xi_t \eta_t - \xi_x \eta_t)^2 + a^2 0 + c^2 0 + ab 0 + bc 0 \\
&=& (b^2 - ac) J^2 > 0
\end{eqnarray}

Dos cosas importantes se hicieron

1. Se redujo la ecuacion a la forma canonica (solo con el termino cruzado en la segunda derivad. $a^*=b^*=0$.
2. El sistema sigue siendo **hiperbolico**. Por que el nuevo discriminante es positivo.

El resultado anterior se puede escribir en forma matricial como

$$\begin{pmatrix} a^* & b^* \\ b^* & c^* \end{pmatrix} =
\begin{pmatrix}
\xi_t & \xi_x \\ \eta_t & \eta_x
\end{pmatrix}  
\begin{pmatrix}
a & b \\ b & c
\end{pmatrix}  
\begin{pmatrix} \xi_t & \eta_t \\ \xi_x  & \eta_x  \end{pmatrix}  
$$
Se deja al estudiante la tarea de verificar esto.
Claramente de esta ultima ecuacion se tiene que
usando $\det (AB) = \det(A) \det(B)$.


$$D^* = D J^2 $$

Este resultado (como $J^2>0$) nos dice que la transoformacion $\xi=\xi(x,t)$ ,
$\eta=\eta(x,t)$ es **invariante hiperbolica** ante el cambio de coordenadas.

Recuerde el punto 1. de arriba. La idea de hacer el cambio de variables es que el sistema se reduzca eliminando los terminos en $a^*, c^*$ (cuadraticos) y dejando solo el termino cruzado $b^*$.
Se deja al estudiante verificar que $a^*=c^*=0$ si $\xi=constante$, $\eta=constante$, que son las caracteristicas..
Que $a^*=0$ y que $b^*=0$ significa que



\begin{eqnarray}
a^* &=& a \xi_t^2 + 2 b  \xi_t \xi_x + c \xi_x^2 = 0 \\
c^* &=& a \eta_t^2 + 2 b \eta_t \eta_x  + c \eta_x^2 = 0
\end{eqnarray}
Podemos unificar estas dos ecuaciones en

$$ a \zeta_t^2 + 2 b \zeta_t \zeta_x  + c \zeta_x^2 = 0 $$
donde $\zeta$ toma los valores de $\xi, \eta$.

Trabajemos un poco esta ecuacion. Dividimos por $\zeta_x^2$.

Como queda

$$a \left ( \frac{\zeta_t}{\zeta_x} \right )^2 + 2b \left ( \frac{\zeta_t}{\zeta_x} \right ) + c = 0 $$

Vamos a asumir que estamos sobre las caracteristicas $\zeta(x,t)=\text{constante}$
Es decir, la diferencial es 0

$$ d \zeta = \zeta_t dt + \zeta_x dx = 0. $$
O sea que

$$ \frac{\zeta_t}{\zeta_x} = - \frac{dx}{dt} $$
y obtenemos

$$a \left ( \frac{dx}{dt} \right )^2 - 2b \left ( \frac{dx}{dt} \right ) + c = 0 $$

Esta cuadratica tiene dos soluciones

$$\frac{dx}{dt} = \lambda_{1,2} = \frac{b \pm \sqrt{b^2 - ac}}{a} \tag{7}$$

Estas soluciones son integradas con las dos constantes de integracion (caracteristicas) $\eta, \xi$. Vamos a ver un ejemplo.
