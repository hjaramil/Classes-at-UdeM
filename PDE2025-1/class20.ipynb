# Solucion a la ecuacion de calor en el intervalo $x \in [-L/2, L/2]$.
**Ejemplo 4.3.2**: Asuma el problema

$$
\begin{cases}
u_t - \alpha u_{xx} = 0   \quad , \quad \text{ecuacion de calor} \\
u(x,0) = \delta(x) \quad , \quad \text{condicion inicial} \\
u(-L/2, t) = 0 \quad , \quad L(L/2, t) = 0 \quad , \quad \text{condiciones de frontera}
\end{cases}
$$

Piense en una barra de calor en el intervalo $[-L/2, L/2]$. Este problema es un poco mas dificil que el problema en $[0, L/2]$.

El $\delta(x)$ se llama delta de Dirac (Paul Dirac). Este lo estudiamos en detalle en la proxima seccion. Solo, por ahora, debemos entender que este delta tiene el siguiente efecto

$$\langle \varphi(x), \delta(x) \rangle = \varphi(0). $$

Usamos los mismos pasos de separacion de variables

$$u(x,t) = F(x) G(t) $$
de donde

$$u_t(x,t) = F(x) G'(t)  \quad , \quad u_{xx}(x,t) = F''(x) G(t) $$

De la ecuacion de calor encontramos

$$F(x) G'(t) = \alpha F''(x) G(t) $$
reorganizando

$$ \frac{F''(x)}{F(x)} = \frac{1}{\alpha} \frac{G'(t)}{G(t)} = k $$

De aca salen dos ecuacions ODE

$$F''(x) = k F(x) \quad, \quad G'(t) = k \alpha G(t). $$

Como hemos hecho antes consideramos 3 casos

1. $k=0$. Entonces $F''(x) = k F(x)=0$.

$$ F(x) = C_1 + C_2 x $$

Aplicamos las condiciones de frontera.
$u(\pm L/2, t) = F(\pm L/2) G(t) = 0$.
Asumiendo que $G(t) \ne 0$.

$$ C_1 - \frac{C_2 L}{2} = 0 $$
$$ C_1 + \frac{C_2 L}{2} = 0 $$
Sumamos estas dos ecuaciones y $2C_1 = 0$ o sea que $C_1 =0$, reemplazando esto en cualquiera de las dos ecuaciones encontramos que $C_2=0$. La solucion es la trivial $u(x,t)=0$. No deseada.

2. $k>0$. Asummimos $k= \lambda^2$, con $\lambda \in \mathbb{R}$. En este caso tenemos las dos ecuaciones

$$F''(x) = \lambda^2 F(x) \quad , \quad G'(t) = \lambda^2 \alpha G(t) $$

Al resolver la primera encontramos $(D- \lambda)(D + \lambda) F=0$ donde $D=d/dx$.
La solucion general es

$$F(x) = C_1 \mathrm{e}^{\lambda x} + C_2 \mathrm{e}^{-\lambda x}. $$
Aplicamos las condiciones de bordo

$$F(x) = C_1 \mathrm{e}^{-\lambda L/2} + C_2 \mathrm{e}^{\lambda L/2}=0. \tag{1} $$
$$F(x) = C_1 \mathrm{e}^{\lambda L/2} + C_2 \mathrm{e}^{-\lambda L/2}=0. \tag{2} $$

Si sumamos estas ecuaciones encontramos que

$$2(C_1 + C_2) \cosh(\lambda L/2) = 0. $$
Como $\cosh(\lambda L/2) \ne 0 $
$C_1 + C_2 = 0$.
De forma que $C_1 = -C_2$ y reemplazamos en la ecuacion



Reemplazmos esto en la (2)

$$C_ 1 \mathrm{e}^{\lambda L/2} - C_1 \mathrm{e}^{-\lambda L/2} = C_1 ( \mathrm{e}^{\lambda L/2} - \mathrm{e}^{-\lambda L/2} = 2 C_1 \sinh (\lambda L/2)=0.$$
De nuevo (tarea para esta semana). Porbar que esto ocurre  solo para $\lambda=0$. O sea $k=0$. Este caso ya lo excluimos.

3. $k<0$. Escribimos $k=-\lambda^2$.
Tenemos la ecuacion $(D^2 + \lambda^2)F =0 $, $D=d/dx$. Las soluciones son complejas.

$$F(x) = C_1 \mathrm{e}^{\mathrm{i} \lambda x} + C_2 \mathrm{e}^{-\mathrm{i} \lambda x} $$

Tarea: Muestre que esto tambien se puede escribir en terminos de ecuaciones trigonometrica.

$$F(x) = C_1 \cos \lambda x + C_2 \sin \lambda x $$

En este momento es donde las condiciones de frontera no ayudan mucho. Veamos

$$C_1 \cos \left ( \lambda \left ( \frac{-L}{2}  \right )   \right )
+ C_2 \sin \left ( \lambda \left ( \frac{-L}{2}  \right )  \right ) =0  \quad , \quad \text{ en  } x=-\frac{L}{2}  $$

$$C_1 \cos \left ( \lambda \left ( \frac{L}{2}  \right )   \right )
+ C_2 \sin \left ( \lambda \left ( \frac{L}{2}  \right )  \right ) =0  \quad , \quad \text{ en  } x=\frac{L}{2}  $$

Resolver esta ecuacion no es facil. La razon por la cual as cosas funcionaron bien antes es por que $\sin 0 = 0$. Entonces vamos a hacer un cambio de variable

$$\hat{x} = x + \frac{L}{2}. $$
Ahora vemos que $\hat{x} = [0, L]$. La ecuacion diferencial de calor no cambia por que $L/2$ es constante.

Lo unico que cambia son las condicines de frontera en la nueva variable $\hat{x}$

$$F(\hat{x}=0) = F(-L/2) = C_1 \cos 0 + C_2 \sin 0  = 0 $$
$$F(\hat{x}=L) = F(L/2) = C_1 \cos \lambda L + C_2 \sin \lambda L  = 0 $$



De la primera de estas dos ecuaciones y $\sin 0=0$ tenemos que $C_1 =0$
La segunda ecuacion se vuelve

$$C_2 \sin \lambda L = 0. $$

La solucion de $\sin x= 0$, es $x=0, \pi, 2 \pi, \cdots, n \pi, \cdots, $.
Si $\lambda L=n\pi$, entonces

$$\lambda = \frac{n \pi}{L} $$
Estos son los **autovalores** del operador SLO.

Cambiemos el nombre para amarrar $\lambda$ con $n$.

$$\lambda_n = \frac{n \pi}{L} $$

Las autofunciones ( o funciones propias) son

$$F_n(\hat{x}) = C_{2n} \sin \frac{n \pi \hat{x}}{L} $$

Escribimos esto en terminos de $x$. Recuerde $\hat{x}=x+L/2$.

$$F_n(x) = C_{2n} \sin \left [ \left ( \frac{n \pi}{L} \right ) \left (x + \frac{L}{2} \right )   \right ] $$
Estas son las autofunciones espaciales. Ahora tomamos la ecuacion en tiempo

$$G'(t) = \alpha \lambda^2 G(t) $$
Esta ecaucion tiene como sulucion


$$G(t) = C_3 \mathrm{e}^{-\alpha \lambda^2 t} $$
Reem[plazmos $\lambda$ por $\lambda_n$ y ponemos un indice $n$ en $G$ o sea $G_n$.

$$G_n(t) = C_{3n} \mathrm{e}^{-\alpha \left ( \frac{n \pi}{L} \right ) ^2 t} $$
Estas son las autofunciones en tiempo.

El producto de ellas es la solucion para el valor $n$.

$$u_n(x,t) = a_n \sin \left [ \left ( \frac{n \pi}{L} \right ) \left ( x + \frac{L}{2}  \right )   \right ] \mathrm{e}^{- \left ( \frac{n \pi}{L}  \right )^2 t}$$
donde $a_n= C_{2n} C_{3n}$.

Hacemos **superposicion**

$$u(x,t) = \sum_{n=1}^{\infty}  a_n \sin \left [ \left ( \frac{n \pi}{L} \right ) \left ( x + \frac{L}{2}  \right )   \right ] \mathrm{e}^{- \left ( \frac{n \pi}{L}  \right )^2 t} $$

Para hallar los coeficientes $a_n$ usamos la condicion inicial.

$$u(x,0) = \delta(x) = \sum_{n=1}^{\infty}  a_n  \sin \left [ \left ( \frac{n \pi}{L} \right ) \left ( x + \frac{L}{2}  \right )   \right ]   $$

Para hallar $a_n$ proyectamos $\delta(x)$ (la condicion inicial) en cada una de las autofunciones

$$ B = \{ \phi_n \} = \left \{  \sin \left [ \left ( \frac{n \pi}{L} \right ) \left ( x + \frac{L}{2}  \right )   \right ]    \right \}.$$

Es decir

$$\langle \delta(x), \phi_k \rangle  = \sum_{n=1}^{\infty} a_n
\left \langle  \sin \left [ \left ( \frac{n \pi}{L} \left ( x + \frac{L}{2} \right )  \right )   \right ] , \sin \left [ \left ( \frac{k \pi}{L} \left ( x + \frac{L}{2} \right )  \right )   \right ]  \right \rangle   \tag{3}  $$
Calculamos esos productos internos


$$\langle \phi_n(x) , \phi_k(x) \rangle =
\int_{-L/2}^{L/2} \sin \left [ \left ( \frac{n \pi}{L} \left ( x + \frac{L}{2} \right )  \right )   \right ]
\sin \left [ \left ( \frac{k \pi}{L} \left ( x + \frac{L}{2} \right )  \right )   \right ]  dx = \delta_{nk} \frac{L}{2}$$

Encontramos de (3) que
$$\delta(x), \phi_k(x) = \sum_{n=1}^{\infty} a_{n} \delta_{nk} \frac{L}{2} = a_k \frac{L}{2}  $$

de donde

$$a_k = \frac{2}{L} \langle \delta(x), \phi_k(x) \rangle = \frac{2}{L} \phi_k(0) = \frac{2}{L} \sin \frac{k \pi}{2} $$

La solucion al final es

$$u(x,t) = \frac{2}{L} \sum_{n=1}^{\infty} \sin \left (  \frac{n \pi}{2} \right )
   \sin \left [ \left ( \frac{n \pi}{L} \right ) \left ( x + \frac{L}{2}  \right )   \right ] \mathrm{e}^{- \left ( \frac{n \pi}{L}  \right )^2 t} $$


# Metodo de separacion de variables en dos dimensiones para la Ecuacion de Laplace.
Abandonamos la variable $t$ (tiempo) y agregamos la variable espacial $y$.

La ecucacion de **Laplace** es:

$$\nabla^2 u(x,y) = 0 $$
donde

$$\nabla^2 \equiv \nabla \cdot \nabla = \frac{\partial^2}{\partial x^2} + \frac{\partial^2}{\partial y} $$

Considermos condiciones de bordo en una caja $[0,a] \times [0, b]$.

Las siguientes condiciones de frontera.

$$
\begin{cases}
u(0, y) = u(a,y) = u(x,0) = 0  \\
u(x,b) = \varphi(x).
\end{cases}
$$

### Separacion de variables

$$u(x,y) = F(x) G(y) $$
De aca

$$ F''(x) G(y ) + F(x) G''(y) = 0 $$
Dividiendo por $F(x) G(y)$,

$$\frac{F''(x)}{F(x)} = -\frac{G''(x)}{G(x)} = k $$

$k$ es constante por los argumentos ya explicados varias veces.
Se originan dos ODE (Ordinary Differential Equations)

$$F''(x) = k F(x) \quad, \quad G''(y) = -k G(y) $$

Como siempre vamos a considerar tres valores de $k$

1. $k=0$: De $F''(x)=0$ tenemos que

$$u(x,y) = F(x) G(y) = (C_1 + C_2 x) G(y) $$
Aplicamos condiciones de bordo

$$ u(0,  y) = 0 \implies C_1 G(y) =0  $$
Como no queremos $G(y)=0$ entonces C_1=0$.
En el bordo derecho

$$ u(a,  y) = 0 \implies C_2 G(y) =0  $$
con el mismo argumento vemos que $C_2=0$. De forma que $F(x)=0$ y la solucion es
$u(x,t)=0$. No nos interesa.

2. $k>0$.
