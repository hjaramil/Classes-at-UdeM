Extendamos el producto interno a funciones de valor complejo. Es decir
$f(x) = a  g(x) + \mathrm{i} h(x) $
donde $g, h: A \subset \mathbb{R} \to \mathbb{R}$ son funciones reales.

En este caso definimos el producto interno como

$$\langle f, g \rangle = \int_D w(x) f(x) \overline{g(x)} dx $$
Se conjunga el $g$. Por ejemplo

$$\langle f, g \rangle = \int_D w(x) f(x) \overline{\lambda g(x)} dx
\langle f, g \rangle = \overline{\lambda} \int_D w(x) f(x) \overline{ g(x)} dx $$

El producto interno tiene las siguientes propiedades. Son faciles de probar

* Distributiva en ambas componentes
$$\langle f + g, h \rangle = \langle f, h \rangle + \langle g, h \rangle  $$
$$\langle f , g + h \rangle = \langle f, g \rangle + \langle f, h \rangle  $$

* Homogenea. Sea $\lambda \in \mathbb{C}$.
$$\langle \lambda f , g \rangle = \lambda \langle f, g \rangle  $$
$$\langle  f , \lambda g \rangle = \overline{\lambda} \langle f, g \rangle  $$

**Definicion 4.1.4** (Norma de una funcion): Sea $f$ una funcion integrable. Entonces
$$\| f \| = \sqrt{\langle f, f \rangle } $$

**Definicion 4.1.5** (ortogonal): dOS FUNCIONES $f,g$ integrables son ortogoanles si

$$\langle f, g \rangle = 0 $$

$$\langle \sin ix , \sin j x \rangle =\int_0^\pi \sin (ix) \sin (jx) dx \tag{3}$$

**Definicion 4.1.6**(operador adjunto): Sea $L$ un operador lineal (diferencial) entonces el **adjunto** de $L$ se define como $L^*$ tal que


$$\langle Lu, v \rangle = \langle u, L^* v \rangle $$
para todo $u$ en el dominio de $L$ y $v$ en el dominio de $L^*$.


Esta definicion es como la extension de la traspuesta de una matriz pero ahora para operadores.

**Ejemplo 4.1.3** (operador adjunto):
Asumamos

$$
\begin{cases}
Lu = \frac{du}{dx}  \quad , \quad x \in [0, 1] \\
u(0) = 2 u(1) \quad , \quad \text{condicion de frontera}
\end{cases}
$$
Encuentre el operador adjunto $L^*$

Los operadores adjuntos diferenciales se hallan por **integracion por partes**.

$$\langle u, v \rangle = \int_0^1 u(x) v(x) dx $$
Entonces

La idea se expresa en la siguiente cadena de ecuaciones.
\begin{eqnarray}
\langle Lu, v \rangle
\\ \vdots  \\
\\ \vdots  \\
\\ \vdots  \\
\\ \vdots  \\
\langle u, L^*v \rangle
\end{eqnarray}

Veamos
\begin{eqnarray}
\langle Lu, v \rangle &=& \int_0^1 u'(x) v(x) dx \\
&=& u(1) v(1) - u(0) v(0) - \int_0^1 u(x) v'(x) dx \quad , \quad \text{integracion por partes } \\
&=& u(1) v(1) - 2 u(1) v(0) -  \int_0^1 u(x) v'(x) dx \\
&=& u(1) ( v(1) - 2 v(0)) -  \int_0^1 u(x) v'(x) dx \\
&=& u(1) ( v(1) - 2 v(0)) - \langle u, L^* v \rangle
\end{eqnarray}
donde defino

$$L^*v = - \frac{dv}{dx} $$
Es decir

$$\langle Lu, v \rangle = \langle u, L^* v \rangle + u(1) (v(1)- 2 v(0)) $$

Para hacer $\langle Lu, v \rangle = \langle u, L^* v \rangle $ necesitamos que el termino adicional sea 0. Es decir $v(1) = 2 v(0)$.  Entonces obtenemos
el problema **dual**

$$
\begin{cases}
L^* v = -\frac{dv}{dx} \\
v(1) = 2 v(0) \quad , \quad \text{condicion de frontera}
\end{cases}
$$
En general, luego de hacer integracion por partes

$$\langle L u, v \rangle = J(u,v) + \langle u , L^* v \rangle $$
donde $J(u,v)=0$ son las condiciones de frontera. En el ejemplo anterior
$$J(u,v) = u(1) ( v(1) - 2 v(0)) = 0 $$

**Ejemplo 4.1.4** (derivadas de segundo orden).
Considere el siguiente SLP:

$$
\begin{cases}
Lu = a_2 (x) \frac{d^2 u}{dx^2} + a_1(x) \frac{du}{dx} + a_0(x) u \\
\alpha_1 u(a) + \beta_1 \frac{d u(x)}{dx} = 0 \quad , \quad
\alpha_2 u(b) + \beta_2 \frac{du(b)}{dx} = 0 \quad , \quad \text{condiciones de bordo}
\end{cases}
$$
Encuentre el operador **adjunto**.

\begin{eqnarray}
\langle Lu, v \rangle &=& \int_a^b  [a_2(x) u''(x) + a_1(x) u'(x) + a_0(x) u(x)] v(x) \\
&=&  \int_a^b  a_2(x) v(x) u''(x) + a_1(x) v(x) u'(x) + a_0(x) v(x)  u(x) \\
\end{eqnarray}
La integracion por partes, la vamos a hacer "por partes" (valga la redundancia). Primero trabamos el termino en $u''(x)$.

1. El termino $u''(x)$
\begin{eqnarray}
\int_a^b [ a_2(x) v(x)] u''(x) dx &=& \left . [a_2(x) v(x)] u'(x) \right |_a^b - \int_a^b u'(x) [ a_2(x) v(x)]' dx  \\
&=& \left . [a_2(x) v(x)] u'(x) \right |_a^b -  \left .[a_2(x) v(x)]' u(x) \right |_a^b  + \int_a^b [a_2(x) v(x)]'' u(x) dx \tag{4}
\end{eqnarray}

2. El termino $u'(x)$

$$\int_a^b [a_1(x) v(x)] u'(x) dx = \left . [a_1(x) v(x) u(x)] \right |_a^b - \int_a^b [a_1(x) v(x)]' u(x) dx \tag{5} $$

3. El termino en $u(x)$ no hay que tocarlo.

$$
\int_a^b a_0(x) v(x) u(x) dx   =   \int_a^b a_0(x) v(x) u(x) dx  \tag{6}
$$
Ahora de (4)+(5)+(5)

$$\langle Lu, v \rangle = J(u,v) + \int_a^b [ [a_2(x) v(x)]'' -[a_1(x) v(x)]' + a_0(x) ] u(x)  = J(u,v ) + \langle u, L^* v \rangle. $$
Donde

$$J(u,v ) =   \left .[a_2(x) v(x)]' u(x) \right |_a^b  +  \left . [a_1(x) v(x) u(x)] \right |_a^b $$
Conclusion el operador adjunto, junto con sus condiciones de frontera

$$ \begin{cases}
L^* v = \frac{d^2 }{dx^2}[ a_2(x) v(x)] - \frac{d}{dx} [ a_1(x) v(x)] + a_0(x) v(x) \\
J(u,v ) = 0  \quad , \quad \text{condiciones de borde}
\end{cases} $$

Por que es importante que un operador sea auto-adjunto? Lo vemos en el siguiente teorema.

**Teorema 4.1.1** (teorema espectral de operadores autoadjuntos): Si $L$ es autoadjunto sus autovalores son **reales** y las autofunciones son **ortogonales**.

**Prueba**: Asuma que $\lambda_j$ es un autovalore de $L$ con autovector $u_j$.
Es decir $L u_j = \lambda_j u_j$.  Sin perdida de generalidad podemos asumir que $\| u_j \|=1$.

1. Primero
Veamos ahora la siguiente forma cuadratica

$$\langle L u_j, u_j \rangle = \langle u_j, L^* u_j \rangle = \langle u_j , L u_j \rangle = \langle u_, \lambda_j u_j \rangle = \overline{\lambda_j} \langle u_j , u_j \rangle =  \overline{\lambda_j}  \tag{7} $$

De otro lado

$$\langle L u_j , u_j \rangle = \langle \lambda u_j, u_j  \rangle = \lambda_j \langle u_j, u_j \rangle = \lambda_j \tag{8} $$

De (7)=(8) tenemos que $\lambda_j = \overline{\lambda_j}$.

Si un complejo es igual a su conjugado, el numero es real: Si $a,b \in \mathbb{R}$

$$a + \mathrm{i} b = a - \mathrm{i} b \implies 2 \mathrm{i} b = 0 \implies b = 0  $$
O sea que $\lambda_j \in \mathbb{R}$.
