## Metodo de las caracteristicas
### Historia.

El metodo de las caracteristicas fue introducido por Gaspard Monge (1746-1818) y Augustin-Louis Cauchy (1789-1857), y Joseph Liouvielle (1809-1882). Sin embargo hay autores que dicen otra cosa. Los autores Yehuda Pinchover y Jacob Rubinstein les dan los meritos a William Hamilton, Christian Huygens, Tomas Young,Augustin Fresnel y Pierre de Ferman. No voy a mencionar mas el asunto.


Una superficie puede interpretarse como un conjunto de curvas especiales llamadas características. La peculiaridad de estas curvas es que, a lo largo de ellas, una ecuación diferencial parcial se transforma en una ecuación diferencial ordinaria (en una sola variable). A continuación, exploramos este método, que introducimos a través de un ejemplo.


Para introducir el metodo de las caracteristica vamos a usar la ecuacion de adveccion

$$ \frac{\partial u}{\partial x} + \frac{\partial u}{\partial t} = 0 \tag{6} $$

Inicialmente definimos el vector de coeficientes de la ecuacion

$$\bma = \binom{1}{1}. $$

Con esto podemos escribir la ecuacion (6) como

$$\bma \cdot \nabla u = 0 $$
Aca $\nabla$ es gradiente

De forma que $a$ es perpendicular al gradiente $\nabla u$ y por lo tanto paralelo a las curvas de nivel definidas como $u(x,t)=constante$. Si $C$ es una curva de nivel, el campo a lo largo de esta curva tiene derivada 0.

Regla de la cadena

$$\frac{du}{ds} = \frac{\partial u}{\partial x} \frac{dx}{ds} + \frac{\partial u}{\partial t} \frac{dt}{ds} = 0 $$

Entonces si definimos el vector tangente $\bmt$


$$\bmt = \binom{ \frac{dx}{ds} }{ \frac{dt}{ds}} $$
tambien es perpendicular al gradiente, pues
$\bmt \cdot \nabla u = 0$.


Los vectores $\bmt,a$ son paralelos de forma ques $\bmt=ka$ o sea

$$ \frac{dx}{ds} = k (1) $$
$$ \frac{dt}{ds} = k (1) $$

El vector tangente usando longitud de arco $s$ es unitario de forma que
$dx/ds, dt/ds$ no pueden desvancer simultaneamente y para $k \ne 0$


$$dx/dt = (dx/ds)(ds/dt) = (dx/ds)/(dt/ds)  $$
pues $ds/dt = 1/(dt/ds)$
$$\frac{dx}{dt} = \frac{dx/ds}{dt/ds} $$


De aca que

$$\frac{dx}{dt} = \frac{k}{k}= 1 $$
y obtenemos la ecuacion diferencial **ordinaria**

$$\frac{dx}{dt} = 1 $$
Cuya solucion es

$$x = t+ c $$

Veamos un ejemplo para funciones de $\mathbb{R}^2 \to \mathbb{R}^2$

Definamos el sistema lineal

\begin{eqnarray}
ax + b y &=& u  \\
cx + dy &=& v
\end{eqnarray}
Podemos escribir este sistema como dos ecuaciones

$$f_1(x,y) = ax + by - u = 0 \tag{1} $$
$$f_2(x,y) = cx + dy - v = 0  \tag{2} $$

Pensamos esto como un sistema de docs ecuacions con dos incognitas $(x,y)$.

El sistema tiene un **Jacobiano** (matriz de derivadas parciales)

$$J_M = \begin{pmatrix}
\frac{\partial f_1}{\partial x} & \frac{\partial f_1}{\partial y} \\
\frac{\partial f_2}{\partial x} & \frac{\partial f_2}{\partial y} \\
\end{pmatrix}= \begin{pmatrix} a & b \\ c& d \end{pmatrix} = A $$

De forma matricial, el sistema (1) (2) se puede escribir como

$$\bmy = A \bmx $$
donde $\bmx = \binom{x}{y} $, y $\bmy=\binom{f_1}{f_2}$.

Este sistema tiene inverso si $A^{-1}$ existe, es decir si el determinante de $A$ no es 0.

Para que exista solucion se requiere
$$\det A \ne 0  $$
Es decir que el Jacobiano $\det A \ne 0$.

Pero los sistemas no tiene que ser lineales.  Por ejemplo


$$f(x,y) = \binom{x^2}{xy} = \binom{u}{v} $$
Dejamos como ejercicio al estudiante encontrar $x,y$ como funciones de $u,v$.

Una razon por la cual el jacobiano no debe ser 0 es por que en aquellos puntos (y pensemos en funciones de $\mathbb{R} \to \mathbb{R}$ donde la derivada es 0, entonces la inversa tiene derivada $1/0 \infty $.

Ejemplo

$$  f(x) = \frac12 x^2 = y  $$

$$ x = \sqrt{2y}$$

$$ \frac{dx}{dy} = \frac{\sqrt{2}}{\sqrt{2y}}$$
El punto donde la derivada de $f$ es 0 es $(0,0)$, pero si $y=0$ esta expresion es $\infty$.

En general, asuma un sistema lineal de la forma $A \bmx =\bmy$, donde
$A$ es una matriz real $n \times n$.


$$\sum_{j=1}^n a_{ij} x_i = y_i \quad , \quad x_i, y_i \in \mathbb{R} \quad , \quad i=1,2, \cdots, n  \tag{3}$$

Este sistema se puede escribir como

$$f_i(x_1, x_2, \cdots, x_n) = \sum_{j=1}^n a_{ij} x_i - y_i \quad , \quad i=1,2, \cdots, n $$

Sabemos, del calculo en varias variables que


$$ a_{ij} = \frac{\partial f_i}{\partial x_j} $$
Que el sistema (3) tenga solucion es equivalente a que el determinante de la matriz $A$ no sea cero, es decir que el jacobiano no sea 0.

$$J_M = \left (  \frac{\partial f_i}{\partial x_j} \right ) \quad , i=1, \cdots, n \quad , \quad j=1, \cdots, n $$
Es decir $J = \det J_M \ne 0$.

El sistema $f_i(\bmx)$, $\bmx \in \mathbb{R}^n$, $i=1,2, \cdots, n$
**no tiene que ser lineal**

# Descripcion general del metodo de las caracteristicas.

## Definiciones basicas.

**primer orden quasi-linear**:

$$a(x,y,u) u_x + b(x,y,u) u_y = c(x,y,u) $$
Aca todo es **real**. Se busca $u=u(x,y)$.
La palabra **quasi** viene de que es lineal en las derivas parciales $u_x, u_y$.

**primer orden linear**:
$$a(x,y) u_x + b(x,y) u_y = c(x,y) u + f(x,y) $$
Lineal quiere decir, que es lineal (potencia 1) en $u_x, u_y, u$.

**primer orden semi-linear**:

$$a(x,y) u_x + b(x,y) u_y =   f(x,y,u) $$
Los dos ultimos casos son casos particulares del primero. Nos vamos a enfocar en el primero.



## Ecuaciones lineales parciales de primer orden

donde $c(x,y,u)= f(x,y) - c(x,y) u$, aca $c$ se repite, pero la diferencia es que el de la izquierda tiene tres argumentos y el de la derecha 2.
