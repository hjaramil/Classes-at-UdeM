**Solucion**: Recuerden de la clase anterior que las autofunciones del SLO son

$$u_n(x) = \sin n x \quad , \quad x \in [0, \pi] \quad , \quad n=1,2, \cdots $$
Debemos asumir $i \ne j$ y tomar dos funciones de este espacio
Verificamos que $\langle \sin i x , \sin j x \rangle = 0$.
La unica funcion es que es ortogonal a sisma es $f(x)=0$.
Veamos. Asumimos $w=1$.

$$\langle \sin ix , \sin j x \rangle =\int_0^\pi \sin (ix) \sin (jx) dx \tag{3}$$
Usamos la siguiente identidad trigonometrica.

$$\sin ix \sin jx = \frac12 [ \cos ( (i-j)x) - \cos ( (i+j)x] $$

La integral (3) se resuelve como sigue
$$\langle \sin ix , \sin j x \rangle =\frac12 \int_0^\pi \sin (ix) \sin (jx) dx = \frac12 \int_0^\pi \cos [ (i-j)x] - \cos[ (i+j) x] dx = \left .\frac{\sin[(i-j)x}{i-j} - \frac{\sin[ (i+j) x}{i+j} \right |_0^\pi = 0.$$

Es decir que las autofunciones $\sin nx$ son ortogonales.

**Definicion 4.1.6**(operador adjunto): Sea $L$ un operador lineal (diferencial) entonces el **adjunto** de $L$ se define como $L^*$ tal que


$$\langle Lu, v \rangle = \langle u, L^* v \rangle $$
para todo $u$ en el dominio de $L$ y $v$ en el dominio de $L^*$.


Esta definicion es como la extension de la traspuesta de una matriz pero ahora para operadores.

**Ejemplo 4.1.3** (operador adjunto):
Asumamos

$$
\begin{cases}
Lu = \frac{du}{dx}  \quad , \quad x \in [0, 1] \\
u(0) = 2 u(1) \quad , \quad \text{condicion de frontera}
\end{cases}
$$
Encuentre el operador adjunto $L^*$

Los operadores adjuntos diferenciales se hallan por **integracion por partes**.

$$\langle u, v \rangle = \int_0^1 u(x) v(x) dx $$
Entonces

La idea se expresa en la siguiente cadena de ecuaciones.
\begin{eqnarray}
\langle Lu, v \rangle
\\ \vdots  \\
\\ \vdots  \\
\\ \vdots  \\
\\ \vdots  \\
\langle u, L^*v \rangle
\end{eqnarray}

Veamos
\begin{eqnarray}
\langle Lu, v \rangle &=& \int_0^1 u'(x) v(x) dx \\
&=& u(1) v(1) - u(0) v(0) - \int_0^1 u(x) v'(x) dx \quad , \quad \text{integracion por partes } \\
&=& u(1) v(1) - 2 u(1) v(0) -  \int_0^1 u(x) v'(x) dx \\
&=& u(1) ( v(1) - 2 v(0)) -  \int_0^1 u(x) v'(x) dx \\
&=& u(1) ( v(1) - 2 v(0)) - \langle u, L^* v \rangle
\end{eqnarray}
donde defino

$$L^*v = - \frac{dv}{dx} $$
Es decir

$$\langle Lu, v \rangle = \langle u, L^* v \rangle + u(1) (v(1)- 2 v(0)) $$

Para hacer $\langle Lu, v \rangle = \langle u, L^* v \rangle $ necesitamos que el termino adicional sea 0. Es decir $v(1) = 2 v(0)$.  Entonces obtenemos
el problema **dual**

$$
\begin{cases}
L^* v = -\frac{dv}{dx} \\
v(1) = 2 v(0) \quad , \quad \text{condicion de frontera}
\end{cases}
$$
En general, luego de hacer integracion por partes

$$\langle L u, v \rangle = J(u,v) + \langle u , L^* v \rangle $$
donde $J(u,v)=0$ son las condiciones de frontera. En el ejemplo anterior
$$J(u,v) = u(1) ( v(1) - 2 v(0)) = 0 $$

Donde

Por que es importante que un operador sea auto-adjunto? Lo vemos en el siguiente teorema.

**Teorema 4.1.1** (teorema espectral de operadores autoadjuntos): Si $L$ es autoadjunto sus autovalores son **reales** y las autofunciones son **ortogonales**.

**Prueba**: Asuma que $\lambda_j$ es un autovalore de $L$ con autovector $u_j$.

Veamos ahora la siguiente forma cuadratica

$$\langle L u_j, u_j \rangle = \langle u_j, L^* u_j \rangle = \langle u_j , L u_j \rangle = \langle u_, \lambda_j u_j \rangle = \overline{\lambda_j} \langle u_j , u_j \rangle =  \overline{\lambda_j}  \tag{7} $$

De otro lado

$$\langle L u_j , u_j \rangle = \langle \lambda u_j, u_j  \rangle = \lambda_j \langle u_j, u_j \rangle = \lambda_j \tag{8} $$

De (7)=(8) tenemos que $\lambda_j = \overline{\lambda_j}$.

Si un complejo es igual a su conjugado, el numero es real: Si $a,b \in \mathbb{R}$

$$a + \mathrm{i} b = a - \mathrm{i} b \implies 2 \mathrm{i} b = 0 \implies b = 0  $$
O sea que $\lambda_j \in \mathbb{R}$.

**Por que son importantes las autofunciones ortogonales**?
Pensemos en $\mathbb{R}^2$.
Escojamos la siguiente base canonica

$$\hat{e}_1 = \binom{1}{0} \quad , \quad \hat{e}_2 = \binom{0}{1} $$

Entoces cualquier vector en $\mathbb{R}^2$ se puede escribir como la suma de dos vectores de estos con pesos, donde los pesos son **faciles** de hallar.

Del algebra lineal
el vector $v = \binom{a}{b}$

$$ \binom{a}{b} = a \hat{e}_1 + b \hat{e}_2 = \langle v, e_1 \rangle \hat{e}_1 + \langle v, \hat{e}_2 \rangle  \hat{e}_2 $$
ademas

$$ \langle v,  \hat{e}_1 \rangle = a $$
$$ \langle v,  \hat{e}_2 \rangle = b $$

Asumamos $n$ dimensiones en vez de dos. Podemos hacer exactamente el mismo procedimiento. Es decir $v \in \mathbb{R}^n$

$$v = \begin{pmatrix}
c_1  \\
c_2 \\
\vdots \\
c_n
\end{pmatrix} $$

$$ v = \sum_{j=1}^n c_j \varphi_j $$
donde $\varphi_i$ son $n$ funciones mutuamente ortogonales y con norma 1.
Es decir

$$\langle \varphi_i, \varphi_j \rangle = \delta_{ij} $$
Aca $\delta_{ij}$ es el **Delta de Kronecker** definida como


$$\delta_{ij} =
\begin{cases}
1  \quad , \quad \text{si  } i = j \\
0  \quad , \quad \text{si  } i \ne j \\
\end{cases}$$
Este $\delta$ tiene la propiedad de "tumbar" (eliminar) una suma. Por ejemplo

$$\sum_{j=1}^n a_j \delta_{ij} = a_i \tag{1} $$

El hecho de que las $\varphi_i$ son ortonormales nos muestra que los coeficientes $c_i$ son la proyeccion de $v$ en cada uno de los miembros de la base $\varphi_i$.

$$\langle v, \varphi_i \rangle = \sum_{j=1}^n  c_j \langle \varphi_j, \varphi_i \rangle  = \sum_{j=1}^n c_j \delta_{ij} = c_i$$

Es decir ,
$$ v = \sum_{i=1}^n \langle v, \varphi_i \rangle \varphi_i  $$

Puedo extender este problema al infinito, de igual forma. En el analisis anterior solo cambien $n$ por $\infty$. Obtenemos

$$v = \sum_{i=1}^{\infty} \langle v , \varphi_i \rangle \varphi_i \tag{2} $$

Los coeficientes, proyeccion de $v$ en cada una de las bases

$$c_i = \langle v, \varphi_i \rangle $$
se llaman **Fourier coefficients** y la expresion (2) es una **serie de Fourier**.

Definicion de base **completo**:

**Definicion 4.1.9** (base completo):  La base $B=\{ \varphi_i \}_{i=1}^{\infty} $ se denomina completa  si para todo vector $v$ en el espacio
(el espacio en que trabajamos es $L^2[a,b]$ (funciones cuyo cuadrado es integrable en el intervalo $[a,b]$ (la "$L$" significa Lebesgue pero aca podemos pensar en Riemann.) que trabajamos cualquier vector $v$ se puede escribir como la suma


$$v = \sum_{i=1}^{\infty} \langle v , \varphi_i \rangle \varphi_i  $$
en terminos de la base.

## Metodo de separacion de variables para la ecuacion de onda unidimensional.

### Motivacion:
Miremos la Figura 4.1

image.png

Si quisieramos usar el metodo de d'Alembert y estamos por alla lejos, como en $1000 \ell$, necesitariamos como 100 ramas y se hace inmanejable y obvio es imposible escribir la solucion en el intervalo $(\infty, \infty)$ usando ramas de d'Alembert (necesitaria un numero infinito de ramas). Entonces el metodo de separacion de variables viene a salvarnos de este problema.

La **receta** es la siguiente

1. Asuma $u(x,t)=F(x) G(t)$. Es decir que $u(x,t)$ es el producto de dos funciones, una en espacio y otra en tiempo.

2. Inserte esta solucion en la ecuacion original (PDE). En este caso la ecuacion de onda. Esto le produce 2 ODE (Ordinary Differential Equations) ecuaciones ordinarias. Que puede resolver mas facilmente.

3. Resuelva las dos ODEs del paso 2 y aplicamos las condiciones de bordo (frontera). Con esto se producen un conjunto infinito de auto-funciones. (son bases del espacio)

4. Use el principio de superposicion para hallar una solucion mas general.
Esto es una suma al infinito (una serie) con coeficientes indeterminados.

5. Con las condiciones iniciales se halla el varlor de los coeficientes indeterminados en el paso 4 y listo.

Veamos un ejemplo

**Ejemplo 4.2.1**: Considere el problema

$$
\begin{cases}
u_{tt} - c^2 u_{xx} = 0  \\
u(0,t) = u(\ell, t) = 0 \quad , \quad \text{condiciones de borde} \\
u(x,0) = \varphi(x) \quad, \quad u_t(x,0) = \psi(x) = 0 \quad, \quad \text{condiciones iniciales}
\end{cases}
$$

Usar el metodo de separacion de variables
1. Asumimos
$$u(x,t) = F(x) G(t)  $$

2.
$$u_{tt} = F(x) G''(t)  \quad , \quad u_{xx} = F''(x) G(t) $$
entonces, de la ecuacion de onda

$$F(x) G''(t) = c^2 F''(x) G(t) $$
De aca que

$$ \frac{F''(x)}{F(x)} = \frac{1}{c^2} \frac{G''(t)}{G(t)} = k $$
$k$ tiene que ser constante por que es imposible que dos funciones sean a la vez funciones de $x$ y de $t$ a no ser que sea una constante.

De forma

$$F''(x) = k F(x) \tag{3} $$
$$ G''(t) = k c^2 G(t). \tag{4} $$
Estas son las 2 ODE del paso 2.

3. Resolvamos la ecuacion (3) $F''(x) = k F(x)$.
