# Funcion de Green de  la ecuacion de Calor.
Asumamos que para $t \ge 0$, $x \in \mathbb{R} queremos resolver el siguiente
IBVP.

$$
\begin{cases}
L u = u_t - \alpha u_{xx} = F(x,t) \\
u(x,0) = 0 \quad , \quad \text{condiciones iniciales homogeneas} \\
u(\pm \infty, t) = 0 \quad , \quad \text{condiciones de frontera homogeneas}
\end{cases}
$$

Vamos a asumir un problema alternativo. Es decir la fuente la vamos a reemplazar por deltas de Dirac y la solucion por la funcion de Green $G(x,t; \xi, \tau)$ puntual en $x=\xi$, y tiempo $t=\tau$. Es decir

$$G_t = \alpha G_{xx} + \delta(x- \xi) \delta(t - \tau). \tag{1} $$

Para resolver esta funcion de Green, vamos a tomar la transforma de Fourier de $x$ a $k$

Recuerden

$$\mathcal{F}[G(x,t; \xi, \tau)] = \hat{G}(k, t; \xi, \tau) = \frac{1}{2 \pi} \int_{-\infty}^\infty G(x, t, \xi, \tau) \mathrm{e}^{-\mathrm{i} k x} dx   $$

Sacando la transformada de Fourier de (1) obtenemos

$$\hat{G_t} = -\alpha k^2 \hat{G} + \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} \delta(t - \tau) \tag{2} $$

Resolvemos la parte homogenea. Es decir estamos en un tiempo $t \ne \tau$
tenemos

$$\hat{G}_t + \alpha k^2 \hat{G} = 0 .$$
Esta ecuacion es ordinaria de primer orden. Facil

$$\hat{G}(k, t; \xi, \tau) = C(k, \xi, \tau) \mathrm{e}^{-\alpha k^2(t - \tau)}. \tag{3} $$

Necesitamos hallar $C(k, \xi, \tau)$

Vamos a integrar la Ecuacion (2) en un intervalo muy pequenho cercano a "la explosion" es decir cercano a $\tau$. El intervalo es
$(\tau -\epsilon, \tau+\epsilon)$.


$$\int_{\tau - \epsilon}^{\tau + \epsilon}  \frac{\partial \hat{G}(k, t; \xi, \tau)}{\partial t} dt = -\alpha k^2 \int_{\tau-\epsilon}^{\tau + \epsilon} \hat{G}(k, t; \xi, \tau) d t + \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} \delta(t - \tau) + \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} $$

Encontramos lo siguiente

$$ \hat{G}(k, t=\tau^+; \xi) - \hat{G}(k, t=\tau^-; \xi) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}. $$
Por que no aparece el termino en la mitad? Vamos a tomar el limite cuando $\epsilon \to 0$. Asumimos que $\hat{G}$ es continua (ver funcion (3)) y por lo tanto la integral entre $-\tau, \tau$ desparece. Ahora bien, por causalidad no existe onda antes de $t = \tau$, entonces la funcion de Green en $\tau^=$ es 0 y tenemos

$$ \hat{G}(k, t=\tau^+; \xi) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}. $$

O sea que en quen $t = \tau$ encontramos de la Ecuacion (3)

$$C(k, \xi, \tau) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} $$

y reemplazando la $C$ en la (3)
$$\hat{G}(x, t; \xi, \tau) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}
\mathrm{e}^{-\alpha k^2(t - \tau)} d \tau $$

Pero esta funcion de Green esta en el dominio del numero de onda $k$. Para pasar al espacio $x$ necesitamos la transformada inversa de Fourier

$$G(x, t ; \xi , \tau) = \int_{-\infty}^\infty \hat{G}(k, t, \xi, \tau) \mathrm{e}^{\mathrm{i}k x} d k = \frac{1}{2 \pi} \int_{-\infty}^\infty \mathrm{e}^{\mathrm{i} k(x - \xi)} \mathrm{e}^{-\alpha k^2(t - \tau)} dk.  $$

Para calcular esta integral se necesitan metodos especiales. El Apendice C muestra que

$$\int_{-\infty}^\infty \mathrm{e}^{\mathrm{i} k b} \mathrm{e}^{-a b^2} dk
= \sqrt{ \frac{\pi}{a}} \mathrm{e}^{- \frac{b^2}{4 a}} $$

Aplicamos esta indentidad con $a = \alpha(t - \tau)$, $b=x - \xi$,
y encontramos

$$G(x, t; \xi , \tau) = \frac{1}{2 \pi} \sqrt{ \frac{\pi}{\alpha(t-\tau)}} \mathrm{e}^{- \frac{(x - \xi)^2}{4 \alpha ( t - \tau) }} $$

Es decir

$$ G(x, t; \xi , \tau) = \frac{1}{4 \pi \alpha(t - \tau) }
\exp \left ( - \frac{(x-\xi)^2}{4 \pi ( t - \tau)} \right )  $$

Ahora bien la solucion es causal. Quiere decir que antes de $\tau$ no existe la onda. Muliplicamos por $H(t - \tau)$ y obtenemos

$$G(x, t; \xi, \tau) =   \frac{1}{4 \pi \alpha(t - \tau) }
\exp \left ( - \frac{(x-\xi)^2}{4 \pi ( t - \tau)} \right ) H(t - \tau). $$

Con esta funcion podemos resolver la ecuacion de calor con fuente. Es decir $Lu = u_t - \alpha u_{xx} + f(x)$. La solucion es

$$u(x,t) = \int_{-\infty}^\infty \int_{-\infty}^\infty F(x,t)
  \frac{1}{4 \pi \alpha(t - \tau) }
\exp \left ( - \frac{(x-\xi)^2}{4 \pi ( t - \tau)} \right ) H(t - \tau) $$

## La funcion de Green para la Ecuacion de Poisson
La ecuacion de Poisson esta dada por

$$\nabla^2 u (x, y) = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2}  = F(x,y) \tag{6} $$
Para hallar la funcion de Green necesitamos

$$\nabla^2 G(x, y, \xi, \eta) = \delta(x - \xi) \delta(y - \eta). $$
Si hallamos esta funcion de Green entonce la solucion de la Ecuacion (6)
esta dada por

$$u(x,y) = \int_{-\infty}^\infty \int_{-\infty}^\infty G(x, y, \xi,, \eta) u(\xi, \eta) d\xi d \eta. $$


El operador de Laplace es simetrico. Es decir no depende de cualquier rotacion que le haga a la solucion. Tarea.
La solucion de la ecuacion de Laplace ($F=0$), es radial, depende solo de $r$.
$r=\sqrt{x^2 + y^2}$.
La solucion es de la forma $u=u(r)$.

Vamos a usar la regla de la cadena para hallar la ecuacion de Laplace en coodenadas cilindricas.

$$\frac{\partial r}{\partial x} = \frac{\partial \sqrt{x^2 + y^2}}{\partial x}  = \frac{2x}{2\sqrt{x^2 + y^2}} = \frac{x}{r}  $$

De igual forma
$$\frac{\partial r}{\partial y} = \frac{y}{r} $$
$r \ne 0 $.
Ahora entonces usando la regla de la cadena

$$\frac{\partial u}{\partial x} = u'(r) \frac{x}{r} \quad , \quad \frac{\partial u}{\partial y} = u'(r) \frac{y}{r}. $$

Ahora calculemos la segundas derivadas.

$$\frac{\partial^2 u}{\partial x^2}= u''(r) \frac{x^2}{r^2} + u'(r)
(x r^{-1})' = u''(r) \frac{x^2}{r^2} + u'(r) \left ( r^{-1} -xr^{-2} \frac{x}{r} \right ) = u''(r) \frac{x^2}{r^2} + u'(r) \left (  \frac{1}{r} - \frac{x^2}{r^3} \right )  \tag{7} $$

De la misma forma encontramos que

$$\frac{\partial^2 u}{\partial y^2} = u''(r) \frac{y^2}{r^2} + u'(r)
\left (  \frac1r - \frac{y^2}{r^3} \right )  \tag{8} $$





Ahora sumamos las  Ecuaciones (7) y la (8) y encontramos que

$$\nabla^2 u(r) = u''(r) + \frac{u'(r)}{r} = \frac1r [ r u'' + u'(r)] = \frac1r \frac{d}{dr}  \left ( r \frac{du}{dr} \right )   $$

Ahora, debemos resolver la ecuacion fundamental
(con fuente en $r_0$, es decir cambiando $r$ a $r - r_0$.

$$\nabla^2 G(r, r_0) = \frac{1}{r - r_0} \frac{d}{dr} \left ( (r-r_0) \frac{du}{dr} \right )  = \delta(r - r_0) \quad , r \ne r_0 \tag{7} $$

Resolvamos primero la homogenea.

$$\nabla^2 G(r, r_0) = \frac{1}{r - r_0} \frac{d}{dr} \left ( (r-r_0) \frac{du}{dr} \right )  = 0 $$.

\begin{eqnarray}
\frac{d}{dr} \left [
    (r - r_0) \frac{d G}{dr}
    \right ] &=& 0  \\
    (r - r_0) \frac{d G}{dr} &=& C_1  \quad , \quad \text{constante} \\
    \frac{d G(r, r_0)}{dr} &=& \frac{C_1}{r-r_0} \\
    G(r, r_0) &=& C_1 \ln (r - r_0) + C_2
\end{eqnarray}

El tercer numeral es integrar la Ecuacion (7) sobre un circulo pequenho al rededor de $r_0$.

$$\int_{B_r(r_0)} \nabla^2 G(r, r_0) dA = 1 $$

Usamos el teorema de la divergencia $\nabla^2 G = \nabla \cdot (\nabla G)$

$$\int_{B_{\epsilon}(r_0)} \nabla G \cdot . n ds = 1 $$

Sabemos que

$$\nabla G(r, r_0) = \frac{C_1}{r - r_0} \hat{r}. $$
donde $\hat{r}$ es un vector unitario radil.

Entonces necesitamos hallar

$$ \int_{B_r(r_0)} \frac{C_1}{\epsilon} ds =  C_1 \frac{2 \pi \epsilon}{\epsilon} = 2 \pi C_1$$

De forma que

$$  C_1 = \frac{1}{2 \pi}$$

Como encontramos $C_2$? asumimos la condicion al infinito de 0. Es decir que la solucion es 0 en el infinitmo. De forma que $C_2=0$ y la solucion es

$$G(r, r_0) = \frac{1}{2 \pi} \ln [ r - r_0]. $$
