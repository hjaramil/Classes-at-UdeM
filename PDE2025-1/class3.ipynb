# Descripcion general del metodo de las caracteristicas.

## Definiciones basicas.

**primer orden quasi-linear**:

$$a(x,y,u) u_x + b(x,y,u) u_y = c(x,y,u) $$
Aca todo es **real**. Se busca $u=u(x,y)$.
La palabra **quasi** viene de que es lineal en las derivas parciales $u_x, u_y$.

**primer orden linear**:
$$a(x,y) u_x + b(x,y) u_y = c(x,y) u + f(x,y) $$
Lineal quiere decir, que es lineal (potencia 1) en $u_x, u_y, u$.

**primer orden semi-linear**:

$$a(x,y) u_x + b(x,y) u_y =   f(x,y,u) $$
Los dos ultimos casos son casos particulares del primero. Nos vamos a enfocar en el primero.



## Ecuaciones lineales parciales de primer orden

Resolvemos el problema

$$a(x,y) u_x + b(x,y)u_y  = c(x,y,u)  \tag{1} $$
donde $c(x,y,u)= f(x,y) - c(x,y) u$, aca $c$ se repite, pero la diferencia es que el de la izquierda tiene tres argumentos y el de la derecha .


Vamos a mostrar una receta para resolver esta ecuacion. El argumento que usaremos es geomtrico.
Asumamos $F=u(x,y)-u = 0$. Entonces queremos pensar que $u(x,y)$ es la solucion (solucion integral) del sistema **integral surface**.
Sabemos que el gradiente de $F$ es normal (perpendicular) a la  superficie

$$\nabla F = [u_x, u_y, -1]  $$

La ecuacion diferencial

$$ a(x,y) u_x + b(x,y) u_y - c(x,y,u) = 0 $$
se puede escribir como un producto punto

$$[a(x,y), b(x,y), c(x,y,u)] \cdot [u_x, u_y, -1] = 0  $$
De forma que vemos que el vector

$$ v = [a(x,y), b(x,y), c(x,y,u)] $$
tambien es normal a la superficie $u$.

Veamos esto en un grafica.



image.png

La superficie de la Figura 2.3 se puede ver como una superficie parametrizada por dos parametros $\sigma, \tau $ (calculo en varias variables)
de la forma

$$x = x(\sigma, \tau) \quad , \quad y=y(\sigma, \tau) \quad , \quad u=u(\sigma, \tau)$$

La supeficie obede (EN 3D) a una ecuacion del radio vector $r$


$$ r(\sigma, \tau) = [ x(\sigma, \tau) , y (\sigma, \tau) , u(\sigma, \tau)]$$

Vamos a fijar una de las coordenadas $\tau=\tau_0$ y la dejamos variable. Entonces estamos fijando una de las curvas sobre la superficie.


$$ r(\sigma, \tau_0) = [ x(\sigma) , y (\sigma) , u(\sigma) ]_{\tau = \tau_0}$$

La tangente de esta curva esta definida (CVV)

$$ t = \frac{d r}{d \sigma}_{\tau = \tau_0} =
\left [ \frac{dx}{d\sigma}, \frac{dy}{d \sigma}, \frac{du}{d \sigma}  \right ] $$

Vamos a escoger, en el punto $x_0=(x(\sigma, \tau_0), y(\sigma, \tau_0), u(\sigma, \tau_0))$.

La clave el asunto es escoger la curva tal que $v$ es paralelo a $t$.
Veamos la siguiente grafica.

image.png

el vector tangente a la curva es $t$, el vector $v$ esta en el plano tangente. Yo puedo escoger el vector $v$ de forma que sea paralelo a $t$ y tendriamos
la ecuacion fundamental

$$t = \lambda v $$

Es decir, componente-a-componente tenemos


$$ \frac{dx}{d \sigma} = \lambda a(x,y) \quad , \quad \frac{dy}{d\sigma} = \lambda b  \quad , \quad \frac{du}{d\sigma} = \lambda c(x,y,u) \tag{2} $$
Estas son las llamadas **ecuaciones caracteristicas** que resuelven la ecuacion diferencial lineal (1)

Observe que lo que hizo fue usar geometrica para **reducir** un sistema de ecuacion diferenciales parciales en un sistema de ecuacion diferenciales ordinarias y este ultimo es mas facil de resolver.


La idea para resolver este problema es resolver las primeras dos ecuaciones para $x  =x(\sigma, \tau_0)$, $y=y(\sigma, \tau)$, luego resolver $u$ como funcion de $\sigma, \tau)$, y finalmente invertir la solucion

$$ x=x(\sigma, \tau) \quad, \quad y = y(\sigma,\tau)$$
de forma que quede

$$\sigma= \sigma(x,y) \quad , \quad \tau=\tau(x,y) $$
e insertar este en $u=(\sigma, \tau)$ para que me quede $u=u(x,y)$.

Formas de solucionar. Podemos dividir la segunda por la primera y la tercera por la primera de las ecuaciones (2)  de forma que


$$ \frac{dy}{dx} = \frac{b(x,y)}{a(x,y)} \quad , \quad \frac{du}{dx} = \frac{c(x,y,u)}{a(x,y)} $$

Podriamos entonces resolver $y=y(x)$ de la primera y $u=u(x)$ de la segunda.
$u=u(x,y(x))$.

**Datos iniciales**
Vamos a asumir, por ejemplo que $\sigma=0$, y la curva para $\tau$ se escribe


$$x = x(0, \tau) \quad, \quad y = y(0, \tau) \quad , \quad u=u(0, \tau) $$

Para cada $\tau$ fijo  resolvemos las caracteristicas y encontramos soluciones

$$ x = x(\sigma, \tau) $$
$$ y = y(\sigma, \tau) $$

Y para poder escribir $\sigma, \tau$ como funciones de $x,y$ debemos **invertir** el sistema. Rerdamos el teorema de la funcion inversa.
Para esto se requiere que el jacobiano no sea 0.
Es decir necesitamos que

$$ J = \begin{vmatrix} x_{\sigma} & x_{\tau}  \\ y_{\sigma} & y_\tau \end{vmatrix}  \ne 0  \tag{3} $$

Vamos a ver por que el decir que $J=0$ es lo mismo que decir que la curva inicial se esta escogiendo tangente a la caracteristica.

Asumamos $J=0$, para $\sigma=0$ (a lo largo de la coordenada $\tau$)
De la ecuacion (3) cambiando $\ne 0$ por $=0$ tenemos

$$ x_\sigma y_\tau - y_\sigma x_\tau = 0 $$
De forma que

$$ \frac{x_\tau}{y_\tau} = \frac{x_\sigma}{y_\sigma} = \frac{a}{b} $$

Estamos asumiento $\sigma=0$ y una curva inicial en funcion de $\tau$.
Es decir la curva $x(\tau), y(\tau)$.  De la tercera ecuacion

$$  \frac{dx}{d \tau} b - \frac{dy}{d \tau} a = 0 \tag{4} $$
Pasando a multiplicar $b$ y $y_\tau$.
El vector tangente en esta curva localmente en $\sigma=0$,

$$t = ( dx/\tau, dy/d \tau) $$. Queremos que este vector tangente no este a lo largo de una caracteristica.  Es decir que no sea tangente  a la caracteristica. La direccion de la caracteristica es $(a,b)$, y la direccion del tangente es $(dx/d \tau, dy/d \tau)$. Matematicamente (algebra lineal) queremos que el determinante de la matriz con las columnas no sea 0.

$$\begin{vmatrix} \frac{dx}{d \tau} & a \\ \frac{dy}{d \tau} & b\end{vmatrix} =\frac{d x}{d \tau} b - \frac{dy}{d \tau} a \ne 0 $$
Es decir contradice la expresion (4). Por lo que necesitamos $J \ne 0$ y esto quiere decir que la caracteristica no es tangente al punto $x_0$ correspondiente a $\sigma=0$.

Mas aun, de la regla de la cadena para derivadas parciales tenemos que

$$u_\tau = u_x x_\tau + u_y y_\tau \quad , \quad \sigma=0 $$
Entonces para $\sigma=0$ tenemos

\begin{eqnarray}
u_\tau &=& y_\tau \left [  \frac{x_\tau}{y_\tau} u_x + u_y  \right ]  \\
&=& y_\tau \left [  \frac{a}{b} u_x + u_y  \right ]  \\
&=& \frac{y_\tau}{b} [ a u_x + b u_y] \\
&=& \frac{y_\tau}{b} c
\end{eqnarray}
Es decir, que en la condicion inicial tenemos de las ecuaciones anteriores
si asumimos que $J=0$ y se cumplen las caracteristicas.

$$ \frac{x_\tau}{a} = \frac{y_\tau}{b} = \frac{u_\tau}{c} $$.
Estas son las ecuaciones de las caracteristicas. O sea que la solucion inicial coincide con la caracteristica si $J=0$.
Esta ultima ecuacion se puede escribir como

$$ \frac{dx}{a} = \frac{dy}{b} = \frac{du}{c} $$
usando diferenciales en vez de derivadas. Es decir en vez de
$dx/d\sigma$ pensamos en $dx$,
$dy/d\sigma$ pensamos en $dy$
$du/d\sigma$ pensamos en $du$
(es como multiplicar todo por $d \sigma$ .

En general pensemos en un vector $t$ tangente a la caracteristica y a los datos iniciales simultaneamente (esto no es bueno, no resuelve el problema)

$$t = \left ( \frac{dx}{d \sigma} , \frac{dy}{d \sigma},  \frac{du}{d \sigma}   \right )  $$
La idea es que $[a,b,c]$ no sea paralelo a $t$.

**Ejemplo 2.2.1**: Este ejemplo tomado de Bleistein
Considere

$$ u_x + u_y = u  $$
$a=1, b=1, c=u$

Las caracteristica son

$$ \frac{dx}{1} = \frac{dy}{1} = \frac{du}{u} $$

El algoritmo es:

1. De las dos primeras encontrar $y = y(x)$
2. De la primera y la tercera encontrar $u=u(x)$.


Paso 1. Dividamos las dos primeras

$$ \frac{dy}{dx} = 1 $$
Solucion $y= x + c_1$.
Dividamos la tercera por la primera

$$ \frac{du}{dx} = u  $$
Solucion
$$ u(x) = c_2 \mathrm{e}^x \tag{6}$$
Es decir tenemos

$$y = x + c_1 \quad , \quad u = c_2 \mathrm{e}^x $$
Todavia no sabemos quien es $c_1, c_2$.

Podemos pensar que amarramos $c_2$ a $c_1$ mediante una funcion $f$. Por ejemplo

$$c_2 = f(c_1) $$

Es decir, de 6 $c_2 = \mathrm{e}^{-x} u$

$$\mathrm{e}^{-x} u = f(c_1) = f(y-x) $$


$$ u = \mathrm{e}^{x}  f(y-x) $$

Pero aun asi no sabemos quien es $f$? $f$ puede ser muchas cosas.
Toca definir $f$, para $x=0$, $u=f(y)$

Por ejemplo si $u=y^2$, $x=0$ (condicion inicial)
Ya con esto la solucion es

$$ u = \mathrm{e}^x ( y- x)^2 $$
A esta solucion se le conoce (Bleistein) como **envelope condition** (mas adelante estudiamos el concepto de "envelope" (envolvente)
