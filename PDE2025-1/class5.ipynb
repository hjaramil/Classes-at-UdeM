## Ecuaciones quasilineales de primero orden (PDEs)
La ecuacion que queremos resolver es:

$$ a(x,y,u) u_x + b(x,y,u) u_y = c(x,y,u)$$

Si seguimos los mismos pasos que hicimos cuando resolvemos la ecuacion lineal obtenemos las mismas caracteristicas pero con coeficientes dependiendo de $u$.

Es decir, recuerden que se crearon dos vectores , uno tangenge $t$ y uno con con los coeficientes $v = [a(x,y,u), b(x,y,u), c(x,y,u)]$ que son perpendicuales al gradiente y por lo tanto en el plano tangente. Se escoje la curva  caracteristica a lo largo de la tangente.  Tenemos por definicion de paralelismo en algebra lineal $t = \lambda v$. Es decir

$$ \frac{dx}{d \sigma} = \lambda a(x,y, u)  
\quad , \quad  \frac{dy}{d \sigma} = \lambda b(x,y, u) \quad , \quad
\frac{du }{d \sigma} = \lambda c(x,y, u)  $$


Algunas veces se escribe de forma simplificada

$$\frac{dx}{a} = \frac{dy}{b} = \frac{du}{c} $$
y si la solucion se puede integrar, dividiendo (para evitar parametrizacion)

$$\frac{dy}{dx} = \frac{b(x,y,u)}{a(x,y,u)} \quad , \frac{du}{dx} = \frac{c(x,y,u)}{a(x,y,u)}. $$

Proxima clase: como usar los datos (curva inicial, parametrizacion en terminos de $\sigma$ y $\tau$. ).

## Especificacion de datos
Asumiendo que la superficie inicialemente esta parametrizada por los parametros $\sigma, \tau$, y que la curva de datos (oblicua a las caracteristicas)
tiene un valor inicial en $\sigma=0$ y esta dada por

$$x = x(0, \tau) \quad , \quad y = y(0, \tau) \quad , u = u(0, \tau). $$


Queremos resolver inicialmente el problema

$$x = x(\sigma, \tau) \quad , \quad y = y(\sigma, \tau) \quad , u = u(\sigma, \tau). $$

Recuerden que esto nos da

$$ x = x ( \sigma, \tau) $$
$$ y = y ( \sigma, \tau) $$

Las debemos **invertir** para hallar $\sigma=\sigma(x,y)$ y $\tau=\tau(x,y)$ y reemplazar esto en $u=u(\sigma(x,y), \tau(x,y))= u(x,y)$.

Para poder invertir este sistema necesitamos la condicion de que el jacobiano no sea 0, es decir


$$J = \begin{vmatrix} x_{\sigma} & x_\tau \\ y_\sigma & y_\tau  \end{vmatrix}  \ne 0 .$$

Otra forma de ver la condicion oblicua (**transversal condition**) es que $t$ no sea paralelo a $v$, es decir

$$t \times v \ne 0  $$
Es decir

$$\begin{vmatrix} e_1 & e_2 & e_3 \\ \frac{dx}{d \tau} & \frac{dy}{d\tau} & \frac{du}{d\tau} \\ a & b & c \end{vmatrix} =
\left [ c \frac{dy}{d \tau} - b \frac{du}{d \tau} , a \frac{du}{d \tau} - c \frac{dx}{d \tau}, b \frac{dx}{d\tau} - a \frac{dy}{d \tau} \right ]  \ne 0 . $$

## Envolvemente de una familia de curvas.
Pensemos en una familia de curvas etiquetada con $\alpha$. Es decir

$$y =f(x; \alpha) $$
Si se puede escribir explicitamente. Ahora veremos el caso mas general donde no hay representacion explita.
La envolvente es una curva que cumple con dos condiciones. Cada punto de la envolvente esta en una curva de la familia, y en ese punto hay una condicion de tangencialidad. Es decir, en ese punto una curva de la familia es tangente a la envolvente. Matematicamente, el punto $(x,y)$ esta en la envolvente si

$$\begin{cases} y = f(x ; \alpha) \quad , \quad \text{para algun  } \alpha \\ \frac{\partial y}{\partial \alpha } = 0 \quad , \quad \text{tangencialidad}.    \end{cases} $$

La idea es eliminar el $\alpha$ de estas dos ecuaciones y hallar $y=y_{envolvente}(x)$

De forma mas general, no asumiendo una representacion explicita podemos escrinbir en vez de $y=f(x; \alpha)$ una funcion general

$$F=F(x,y(x), \alpha)=0 $$
Por ejemplo:

$$ x + 3y + \alpha = 0, \quad , \quad \text{en vez de   }  y = -\frac{\alpha + x}{3}$$

Vamos a ilustrar por que se necesitan dos condiciones

$$\begin{cases} F(x,y( \alpha))= 0 \quad , \quad \text{que el punto esta en la "superficie" $F$}\\ \frac{\partial F}{\partial \alpha} = 0
\quad , \quad \text{condicion de tangencialidad} \end{cases} $$

Nos referimos a la siguiente grafica.



image.png

Escojamos una curva $\alpha_0$ (la verde en este caso). Tenemos que la familia de curvas se denomina como $F(x, y, \alpha_0)$. Llamemos $y_{curv}(x)$ a la curva $\alpha_0$, y llamamos $y_{envelope}$ a la envolvente.

Usamos la regla de la cadena, pensando en $F$ como funcion solo de $x$

\require{cancel}

En la envolvente
$$\frac{dF}{dx} = \frac{\partial F}{\partial x} \frac{\partial x}{\partial x}  + \frac{\partial F}{\partial y} \frac{\partial y_{envolvente}}{\partial x} + \frac{\partial F}{\partial \alpha} \frac{d \alpha}{dx}  = 0$$

Ahora pensemos en una curva $\alpha_0$ de la familia, a lo largo de esa curva tenemos $\alpha=constante$, de forma que $dF/d \alpha_0 = 0$,
De nuevo usando la regla de la cadena

$$\frac{dF}{dx} = \frac{\partial F}{\partial x} \frac{\partial x}{\partial x} + \frac{\partial F}{\partial y} \frac{d y_{curva}}{dx}  = 0. $$

Igualando las dos tenemos

$$\frac{\partial F}{\partial y} \frac{d y_{envelope}}{dx} + \left . \frac{\partial F}{\partial \alpha} \frac{d \alpha}{dx} \right |_{\alpha = \alpha_0} = \frac{\partial F}{\partial y} \frac{d y_{curva}}{dx} \tag{1}$$

Como la tangente es la misma para ambas curvas

$$ \frac{d y_{envolvente}}{dx} = \frac{d y_{curva}}{dx}$$
Entonces el termino de la midad en la Ecuacion (1) es 0. Es decir

$$  \left . \frac{\partial F}{\partial \alpha} \frac{d \alpha}{dx} \right |_{\alpha = \alpha_0} = 0$$
donde asumimos $\alpha = \alpha_0(x)$. Para poder resolver el sistema implicito
$F(x,y, \alpha)=0$, necesitamos que la derivada $d \alpha /dx \ne 0$.
De forma que

$$\frac{\partial F}{\partial x} = 0. $$

Lo que tenemos que resolver

$$\begin{cases} F(x,y,\alpha)= 0  \\ \frac{\partial F}{\partial \alpha} = 0. \end{cases} $$

De estas dos ecuaciones debemos eliminar $\alpha$ y queda una ecuacion solo en $x,y$ (implicita o explicita).

**Ejemplo 2.2.6**: Vamos a considerar el ejemplo de la Figura 2.8. Asuma que el punto mas alto en $y$ es el interpto $c-1$, $c>1$. La primera linea la pensamos como desde $(1,0), hasta $(0, c-1)$, la segunda de desde $(2,0)$, hasta $(c-2)$.
Observe que  la suma de los interceptos estan dados por


$$1 + c-1 = c $$
$$2 + c-2 = c $$
$$ \vdots $$
$$n + c-n = c $$

Como tarea les preguntare que hallen las velocidades de deplazamiento vertical y horizontal y veriquen que no son iguales (solo si el angulo es 45 grados), o sea que este experimento no coincide con el deslizamiento de una escalera).

El parametro $\alpha$ es para nosotros la pendiente de la recta, y el intercepto $b$.

Tenemos la familia de curvas (rectas)

$$ y = \alpha x + b \tag{2} $$
De esta ecuacion debemos eliminar $b, \alpha$
Eliminamos $b$ de la conservacion de suma de catetos (interceptos). Hallamos los interceptos en funcion $b$ los summos, despejamos $b$ y lo reemplazamos en la ecuacion anterior (2).

Cuando $x=0$, $y=b$ y de

$$x = \frac{y - b}{\alpha} $$
de forma que cuando $y=0$, $x=-\frac{b}{\alpha}$.
La suma de interceptos es $c$ o sea


$$ b - \frac{b}{\alpha} = c$$
$$ b \left (  1   - \frac{1}{\alpha} \right ) = c$$
Despejamos $b$

$$ b = \frac{c}{1 - 1/\alpha} = \frac{c \alpha}{\alpha -1} $$
Reemplazamos este $b$ en la ecuacion (2).

$$y = \alpha x + \frac{c \alpha}{\alpha - 1}  $$

Para eliminar $\alpha$ usamos la condiciones de **tangencialidad** ($\partial F/\partial \alpha = 0$)
donde
$$F(x,y,\alpha)= y -\alpha x -\frac{c \alpha}{\alpha -1}  $$

$$\frac{\partial F}{\partial \alpha} = -x -\frac{(\alpha -1)c - c \alpha } {(\alpha -1)^2} = - x - \frac{c \alpha + c - c \alpha}{(\alpha-1)^2} = -x + \frac{c}{(\alpha -1)^2} = 0.$$

Despejamos $\alpha$ y la reemplazamos en (2)
$(\alpha -1)^2 = \frac{c}{x}$
De aca dan dos soluciones


$$\alpha = \pm \sqrt{ \frac{c}{x}} + 1 $$

Sustituimos este $\alpha$ en la ecuacion (2).
Tomemos la solucion "+".

$$y = \left ( 1 + \sqrt{\frac{c}{x}}  \right )x + \frac{c (1 + \sqrt{c/x})}{(1 + \sqrt{c/x}) -1 }   $$

$$y = \left ( 1 + \sqrt{\frac{c}{x}}  \right )x + \frac{c (1 + \sqrt{c/x})}{\sqrt{c/x}  }   $$

$$y = \left ( 1 + \sqrt{\frac{c}{x}}  \right )x + c \left ( 1 + \sqrt{\frac{c}{x}} \right ) \sqrt{\frac{x}{c}}$$
Simplificamos

$$ y = x + \sqrt{cx} + \sqrt{cx} + c  $$

$$ y= x + 2 \sqrt{c x} + c $$

Si seguimos el signo "-"

$$y = \left ( 1 - \sqrt{\frac{c}{x}}  \right )x - c \left ( 1 - \sqrt{\frac{c}{x}} \right ) \sqrt{\frac{x}{c}}$$



```python
import numpy as np
import matplotlib.pyplot as plt
c=100

def envelope(x,c):
    return x + c + 2 * np.sqrt(c*x)

def envelope2(x,c):
    return x + c - 2 * np.sqrt(c*x)

domain = np.arange(0, c, 0.01)


# plot labels
plt.xlabel("x")
plt.ylabel("y")

plt.title("Envelope of curve")

plt.plot(domain, envelope(domain,c))
plt.plot(domain, envelope2(domain,c))
plt.show()
```

```python
# no es claro el panorama. Dibujamos la familia

for i in range(1, 100, 2):
    x_values=[i,0]
    y_values=[0,99-i]
    plt.plot(x_values,y_values, 'b', linewidth=1)


# labels
plt.xlabel("x")
plt.ylabel("y")

plt.title("Family of curves")

# set plot limits
plt.xlim([0, 100])
plt.ylim([0, 100])

# show grid
plt.grid(True)

# set aspect ratio equal to 1
plt.gca().set_aspect('equal', adjustable='box')

plt.plot(domain, envelope2(domain,c), color='r', linewidth=2)

plt.show()
```
