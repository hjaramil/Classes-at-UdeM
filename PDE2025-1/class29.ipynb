## Funcion de Green para la ecuacion de onda.

Debemos resolver

$$u_{tt} - c^2 u_{xx} = F(x,t) $$
con condiciones iniciales homogeneas

$$u(x,0) = 0 \quad , \quad u_t(x,0) = 0  $$
y de frontera

$$u(\pm \infty, t) = 0. $$

La funcion de Green se calcula de la ecuacion

$$G_{tt} - c^2 G_{xx} = \delta(x-\xi) \delta(t - \tau) $$
$G= G(x, t; \xi, \tau) $.

Los pasos son
1. Transformada de Fourier de $x$ a $k$.

$$ \hat{G}_{tt} + c^2 k^2 \hat{C} dt = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} \delta(t-\tau). \tag{4} $$

2. Resolvemos la ecuacion homogenea. Es decir $t \ne \tau$.

$$\hat{G}_{tt} + c^2 k^2 \hat{G} = 0 . $$
Esta ecuacion tiene dos soluciones y a la hemos hecho antes.

$$\hat{G}(k, t; \xi, \tau) = A(k, \xi, \tau) \cos(ck(t - \tau)) +
B(k, \xi, \tau) \sin(ck(t - \tau)). \tag{5} $$

Dejo como tarea explicar por que el cambio de $t$ a $t-\tau$ no afecta el resultado de la ecuacion.

3. Imponemos las condiciones iniciales.
De la misma forma que hicimos en la ecuacion de calor, integramos sobre un intervalo pequenho $(\tau-\epsilon, \tau+ \epsilon)$.
Integramos la Ecuacion (4).



$$ \int_{\tau-\epsilon}^{\tau+\epsilon} \hat{G}_{tt} dt + \int_{\tau-\epsilon}^{\tau+ \epsilon} c^2 k^2 \hat{G} dt = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} $$

La segunda integral es 0 en el limite $\lim_{\epsilon \to 0}$ por que $\hat{G}$ es continua.

La primea integral y el ultimo termino producen

$$\hat{G}_t(k , \tau^+) - \hat{G}(k, \tau^-)  = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}. \tag{6}$$
La expresion del medio es 0 por causalidad. Es decir $\hat{G}(k, \tau^-)=0$.
Tenemos encones que

$$ \hat{G}_t(k , t=\tau) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}.$$
Ahora bien de la Ecuacion (5) con $t = \tau$

En $t = \tau$ ocurre que
$$\hat{G}(k, t=\tau; \xi)=0 $$

Reemplazando $t=\tau$ en (5)

$$\hat{G}(k, t=\tau; \xi, ) = A(k, \xi, \tau)   = 0  $$
O sea que $A=0$, y  de la Ecuacion (5)
tomando la derivada y de $A=0$,

$$ \hat{G_t}(k, t=\tau, \xi) =  B(k, \xi, \tau) = B(k, \xi , \tau) c k \cos(0) = B(k \xi, \tau) c k$$.
Se postpne esta prueba para la proxima clase.
De aca y  la Ecuacion (6)

$$B(k, \xi, \tau) = \frac{1}{2 \pi c k} \mathrm{e}^{-\mathrm{i} k \xi}. $$

La solucion, de $(5) $ y como $A=0$.

$$\hat{G}(k, t; \xi, \tau) = \frac{1}{2 \pi ck} \mathrm{e}^{-\mathrm{i} k \xi} \sin (c k ( t- \tau)). $$

El problema es que esta solucion esta en el dominio del numero de onda $k$ y la necesitamos en $x$.

Debemos tomar la **transformada inversa** para regresar de $k$ a $x$.

Es decir

$$G(x, t; \xi, \tau) = \frac{1}{2 \pi c} \int_{-\infty}^\infty \frac{\sin[c k ( t - \tau)]}{k} \mathrm{e}^{\mathrm{i} k(x - \xi)} d k  $$

Esta integral no es trivial. En el Apendice D la resuelvo.
La solucion es

$$G(x, t ; \xi , \tau) = \frac{1}{2c}
\left (
    H \left (x - (\xi - c(t - \tau)) \right )
    - H \left (x - (\xi + c(t - \tau)) \right )
    \right ) $$


A parte de esto la solucion debe ser causal. Es decir que antes $\tau$ no hay nada. Debe ser multiplicada por $H(t - \tau)$. El resultado es entonces


$$G(x, t ; \xi , \tau) = \frac{1}{2c} H(t - \tau)
\left (
    H \left (x - (\xi - c(t - \tau)) \right )
    - H \left (x - (\xi + c(t - \tau)) \right )
    \right ) $$


La interpretacion de esto. Si asumimos que la fuente en tiempo y espacio es 0, tenemos $\xi=0, \tau=0$.

$$G(x, t ; \xi , \tau) = \frac{1}{2c} H(t - \tau)
\left (
    H \left (x  + c t) \right )
    - H \left (x -  ct \right )
    \right ) $$



# Separacion de variables de la ecuacion de calor en dos dimensiones espaciales.

El problema que queremos resolver es

$$
\begin{cases}
u_t - \alpha ( u_{xx} + u_{yy}) = 0 \quad , \quad \text{PDE} \tag{7}\\
u(0, y, t) = u(L_x, y, t) = u(x, 0, t) = u(x, L_y, 0) = 0 \quad , \quad \text{condiciones de frontera} \\
u(x,y,0) = \varphi(x,y)  \quad , \quad \text{condicion inicial}
\end{cases}
$$

Por separacion de variables tenemos

$$u(x,y,t) = X(x) Y(y) T(t).  $$


Sustituimos esto en la ecuacion de Calor (7) y  obtenemos


$$X(x) Y(y) T'(t) = \alpha ( X''(x) Y(y) T(t) + X(x) Y''(y) T(t))  $$

Ahora dividimos por $u(x,y,t)=X(x) Y(y) T(t)$ y obtenemos

$$\frac{T'(t)}{\alpha T(t)}  = \frac{X''(x)}{X(x)} + \frac{Y''(y)}{Y(y)}. \tag{8} $$


Ahora , el termino de la izquierda debe ser constante por que no existe funcion que sea de $t$ igual a funcion que sea de $x,y$ en todas partes,excepto por una constante.

$$\frac{T'(t)}{\alpha T(t)}= -\lambda^2 $$.

De igual forma, si separamos $X(x)$ y $Y(y)$ de la Ecuacion (8) encontramos que

$$\frac{X''(x)}{X(x)} = -\mu^2  \quad , \quad \frac{Y''(y)}{Y(y)} = - \gamma^2  $$

De la Ecuacion (8) tenemos que


$$\mu^2 + \gamma^2 = \lambda^2.  \tag{9} $$
Para reducir variables odemos decir que $\gamma^2 = \lambda^2 - \mu^2$ y escribir

$$\frac{X''(x)}{X(x)} = -\mu^2  \quad , \quad \frac{Y''(y)}{Y(y)} = \mu^2 - \lambda^2   $$

Debemos resolver estas dos ecuaciones ordinarias. Primero por $X(x)$.

$$X''(x) + \mu^2 X(x) = 0. $$
Las soluciones de esto

$$X(x) = C_1 \cos \mu x + C_2 \sin \mu x $$
Ahora aplicamos condiciones de frontera.o
$$u(0, y, t) = X(0) Y(y) T(t) = 0  $$
o sea $X(0)=0$ (para evitar la solucion trivial)

$$X(0) = C_1 = 0, $$
De froma que

$$X(x) = C_2 \sin \mu x.   $$
La otra condicion de frontera es $X(L_x)=0$ (evitando soluciones triviales).

$$ X(L_x) = C_2 \sin \mu L_x = 0$$.
Esto implica un conjunto infinito de soluciones, pues $\mu L_x = m \pi$.
Entonces

$$\mu = \mu_m = \frac{m \pi}{L_x}  $$
conjunto infinito de **autovalres** del operador SLO.

Paremos aca por un momento y vamos a la ecuacion en $Y(x)$.

$$Y''(y) + (\lambda^2 - \mu^2) Y(y) = 0. $$
con $\lambda^2 - \mu^2 = \gamma^2 > 0 $.

La solucion es
$$Y(y) = C_3 \cos \gamma y +  C_4 \sin \gamma y .$$
Usando las condiciones de frontera en $Y$ encontramos que

$$Y(0) = C_3 = 0. $$
y

$$Y(y) = C_4 \sin \gamma. $$
Ahora usando  $Y(L_y)=0$, encontramos que

$$Y(y) = C_4 \sin \gamma L_y = 0 $$
de forma que $\gamma L_y = n \pi$ (un conjunto infinito de autovalores)

$$ \gamma = \frac{n \pi}{L_y}. $$

Ahora bien, de  la Ecuacion (9) encontramos

$$\lambda_{mn}^2 = \bigg ( \frac{m \pi}{L_x} \bigg )^2 + \bigg ( \frac{n \pi}{L_x} \bigg )^2 $$

Retornemos a la ecuacion en tiempo.

$$\frac{T'(t)}{\alpha T(t)} = - \lambda^2 $$
Esta tiene una solucion exponencial

$$T(t) = C_5 \mathrm{e}^{- \alpha \lambda^2 t} $$
es decir

$$T(t) = C_5 \mathrm{e}^{-\alpha \left [ \left ( \frac{m \pi}{L_x} \right)^2+  \left ( \frac{n \pi}{L_y} \right )^2 \right ] t} $$

Ahora si usamos superposicion.

$$u(x,y,t) = \sum_{m=0}^{\infty} \sum_{n=0}^{\infty} a_{mn}
\sin \frac{m \pi x}{L_x} \sin \frac{n \pi y}{L_y}  
 \mathrm{e}^{-\alpha \left [ \left ( \frac{m \pi}{L_x} \right)^2+  \left ( \frac{n \pi}{L_y} \right )^2 \right ] t}
$$

Como hicimos en la clase anterior, definimos las auto-funciones ortogonales

$$\varphi_{mn} = \sin \frac{m \pi x}{L_x}  \cos \frac{n \pi y}{L_y} $$

Ahora la condicion inicial $t=0$ produce

$$u(x,y, t=0) = \varphi(x,y) = \sum_{m=0}^{\infty} \sum_{n=0}^{\infty} a_{mn}
\sin \frac{m \pi x}{L_x} \sin \frac{n \pi y}{L_y}.  \tag{10}  $$

El $a_{mn}$ como lo hicimos en la clase anterior se halla por el hechode que
$\langle \varphi_{mn}, \varphi_{pq} \rangle = \delta_{mp} \delta_{nq} \frac{L_x}{2} \frac{L_y}{2}$.

El metodo
1. Tomamos el producto interno (proyeccion) de $10$ con $\varphi_{pq}$.
2. Los dos deltas de Kronecker, eliminan las funciones y nos dan los coeficientes $a_{pq}$ en terminos de $L_x, L_y$. Cambiamos $p \to m$, $q \to n$ y obtenemos

$$ a_{mn} = \frac{4}{ L_x L_y} \langle \varphi, \varphi_{mn} \rangle =
 \frac{4}{L_x L_y}= \int_{-\infty}^\infty \int_{-\infty}^\infty \varphi(x,y)
 \sin \frac{ m \pi x }{L_x} \sin \frac{n \pi y}{L_y} dx dy$$

La soluciones es entonces:


$$u(x,y,t) = \sum_{m=0}^{\infty} \sum_{n=0}^{\infty} a_{mn}
\sin \frac{m \pi x}{L_x} \sin \frac{n \pi y}{L_y}  
 \mathrm{e}^{-\alpha \left [ \left ( \frac{m \pi}{L_x} \right)^2+  \left ( \frac{n \pi}{L_y} \right )^2 \right ] t}
$$
con

$$ a_{mn} = \frac{4}{ L_x L_y} \langle \varphi, \varphi_{mn} \rangle =
 \frac{4}{L_x L_y}= \int_{-\infty}^\infty \int_{-\infty}^\infty \varphi(x,y)
 \sin \frac{ m \pi x }{L_x} \sin \frac{n \pi y}{L_y} dx dy$$
