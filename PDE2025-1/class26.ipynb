## Transformada de Fourier de la convolucion.
Asumimos que las funciones $f,g$ so integrables in $(-\infty, \infty)$. Definimos la convolucion como

$$(f \ast g)(x) = \int_{-\infty}^\infty f(x-y) g(y) dy  $$

Entonces la transformada de Fourier esta dada por

$$\mathcal{F} [ (f \ast g)(x)] = \frac{1}{2 \pi} \int_{-\infty}^\infty (f \ast g)(x) \mathrm{e}^{-\mathrm{i} k x} dx = \frac{1}{2 \pi} \int_{-\infty}^\infty \int_{-\infty}^\infty [f(x-y) g(y) dy ] \mathrm{e}^{-\mathrm{i} k x } dx $$

Podemos usar la regla de Fubini (asumiendo que las hipotesis se dan).
Conmutar el orden de las funciones dentro de la doble integral

$$\mathcal{F}[ (f \ast g)(x)] = \frac{1}{2\pi} \int_{-\infty}^\infty g(y)
\left ( \int_{-\infty}^\infty f(x-y) \mathrm{e}^{-\mathrm{i} k x}  \right )  dy $$

Hacemos un cambio de variable $u=x-y$, $du=dx$, y
$\mathrm{e}^{-\mathrm{i} k x} = \mathrm{e}^{-\mathrm{i} k(u+y)} $
y $u \in (-\infty, \infty)$

Es decir

\begin{eqnarray}
\mathcal{F}[(f \ast g)(x)] &=& \frac{1}{2 \pi} \int_{-\infty}^\infty g(y) \mathrm{e}^{-\mathcal{i} k y} dy  \int_{-\infty}^{\infty} f(u) \mathrm{e}^{- \mathrm{i} k u } du \\
    &=& 2 \pi \frac{1}{2 \pi} \int_{-\infty}^\infty g(y) \mathrm{e}^{-\mathrm{i} k y} dy \frac{1}{2 \pi} \int_{-\infty}^{\infty} f(u) \mathrm{e}^{-\mathrm{i} k u} du \\
    &=& 2 \pi \mathcal{F}[f(x)] \mathcal{F}[g(x)].
\end{eqnarray}

Entonces en palabras, la transformada de Fourier de la convolucion de $f$ con $g$ (es decir $f \ast g$)  es el producto (salvo un factor $2 \pi$) de las transformadas de Fourier de  $f$ y de $g$.





## Transformada de Fourier de un operador diferencial.
Considere $f_x = \partial f(x)/\partial x$. $f$ puede ser funcion de varias variables pero por el momento nos enfocamos en la variable $x$.

$$\mathcal{F} \left [ \frac{\partial f(x)}{\partial x} \right ]  =
\frac{1}{2 \pi} \int_{-\infty}^\infty \frac{\partial f(x)}{\partial x} \mathrm{e}^{- \mathrm{i} k x } dx = \frac{1}{2 \pi} [ f(x)] \bigg |_{-\infty}^{\infty}  - \frac{1}{2 \pi} \int_{-\infty}^{\infty} (-\mathrm{i}k) f(x) \mathrm{e}^{-\mathrm{i} k x }  dx.   $$

$f(x)$ no es necesariamente una funcion de prueba (test function) pero si asumimos que decae a 0 en el infinito, entonces el termino
$$  \frac{1}{2 \pi} [ f(x)] \bigg |_{-\infty}^{\infty} $$ se hace 0 y tenemos


$$\mathcal{F} \left [ \frac{\partial f(x)}{\partial x}    \right ] = \mathrm{i} k \int_{-\infty}^{\infty} f(x) \mathrm{e}^{-\mathrm{i} k x } dx = \mathrm{i} k \mathcal{F}[ f(x)] $$

En palabras: La transformada de Fouier de un operador diferencial es $\mathrm{i} k$ por la transformada de Fourier de la funcion. Es decir, la transformada de Fourier transforma operadores diferenciales en operadores algebraicos. Sabemos bien que es mucho facil resolver una ecuacion algebraica que una diferencial. He aca parte de la magia de la transformada de Fourier.

Esto lo podriamos ver de otra forma usando distribuciones


$$\mathcal{F}(f) = \left \langle f, \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k x} \right \rangle  $$

De forma que

$$\mathcal{F}[ f'(x)] = \left \langle f' , \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k x}\right \rangle = - \left \langle f, \mathrm{i} k \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k x} \right \rangle = \mathrm{ik} \left \langle f, \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k x} \right \rangle = \mathrm{i} k \mathcal{F}[ (f(x)] .    $$

Vamos a la segunda derivada.

\begin{eqnarray}
\mathcal{F} [ (f''(x)] &=& \left \langle f'',  \frac{1}{2 \pi} \mathrm{e}^{- \mathrm{i} k x } \right  \rangle \\
&=& - \left \langle f',  -\frac{\mathrm{i} k}{2 \pi} \mathrm{e}^{-\mathrm{i} k }  \right \rangle  \\
&=& \mathrm{i} k \left   \langle f' ,  \frac{1}{2 \pi} \mathrm{e}^{- \mathrm{i} k } \right \rangle  \\
&=& \mathrm{i} k \times  \mathrm{i} k  \left \langle  f ,   \frac{1}{2 \pi} \mathrm{e}^{\mathrm{i} k }  \right  \rangle  \\
&=& -k^2 \mathcal{F}[(f(x)]
\end{eqnarray}

De forma que la transformada de Fourier de la doble derivada de una funcion $f$ es el producto de $-k^2$ por la transformada de Fourier de la funcion.


Lo que acabamos de hacer, por induccion, se puede generalizar a cualquier numero de derivadas. Es decir


$$\mathcal{F} [ f^{(n)}] = (\mathrm{i k})^n \mathcal{F}[ f(x)] $$

Esto nos permite generalizar el concepto de derivada a derivadas **fraccionarias**, por ejemplo la derivada media se puede calcular como

$$\frac{d^{1/2} f(x)}{dx^{1/2}} = \mathcal{F}^{-1}
\left [ \sqrt{\mathrm{i} k} \mathcal{F}[(f(x)]   \right ]   $$

En general podemos definir la **derivada q-fraccionaria**  como

$$\frac{d^{q} f(x)}{dx^{q}} = \mathcal{F}^{-1}
\left [ (\mathrm{i} k)^q \mathcal{F}[(f(x)]   \right ]   $$

Estas derivadas fraccionarias se usan en procesamiento de senhales.

## La transformada de Fourier de una constante, de la Heaviside y de la funcion signo (sng)

* **transformada de Fourier de una constante**:
Asuma una constante $A_0$. Para encontrar la transformada de Fourier de esta constante, asumamos que en el dominio del numero de onda tenemos la siguiente distribucion $A_0 \delta(k)$.  Queremos hallar la transformada inversa de esta distribucion. Es decir,

$$\mathcal{F}^{-1}[ A_0 \delta(k)] = \int_{-\infty}^\infty A_0 \delta(k) \mathrm{e}^{\mathrm{i} k x}  dx  = A_0$$

Ahora tomemos la transformada de Fourier de la expresion anterior. Obtenemos

$$ A_0 \delta(k) = \mathcal{F}[A_0] $$
Es decir, la transforma de Fourier de una constante $A_0$ esta dada por

$$A_0 \delta(k). $$

* **transformada de Fourier de la funcion signo "sgn"**.
Para esta transformada, recuerden que


$$\frac{d \text{sgn}}{dx} = 2 \delta(x). $$

Entoncres

$$\mathcal{F} \left [ \frac{d \text{sgn}}{d x}  \right ] = 2 \mathcal{F}[\delta(x)] = 2 \frac{1}{2 \pi} = \frac{1}{\pi} \tag{1} $$

Usando teorema de la tranformada de la derivada

$$\mathcal{F} \left [  \frac{d \text{sgn}(x)}{d x} \right ] = \mathrm{i} k \mathcal{F} [ \text{sgn} (x) ]  \tag{2} $$

Comparando (1) con (2) encontramos que

$$ \mathcal{F}[ \text{sgn}(x)] = \frac{1}{\mathrm{i} \pi k}. $$

La ultima propiedad es la transformada de Fourier de la Heaviside

* **transformada de Fourier de la Heaviside funcion**:

Recuerden
$$H(x) = \frac{\text{sgn}(x) + 1}{2} $$

De forma que

$$\mathcal{F}[H(x)] = \frac12 ( \mathcal{F}[\text{sgn}(x)] + \mathcal{F}[1] ] = \frac{1}{2 \mathrm{i} \pi k} + \frac{\delta(k)}{2}   $$

# Funcion de Green de  la ecuacion de Calor.
Asumamos que para $t \ge 0$, $x \in \mathbb{R} queremos resolver el siguiente
IBVP.

$$
\begin{cases}
L u = u_t - \alpha u_{xx} = F(x,t) \\
u(x,0) = 0 \quad , \quad \text{condiciones iniciales homogeneas} \\
u(\pm \infty, t) = 0 \quad , \quad \text{condiciones de frontera homogeneas}
\end{cases}
$$

Vamos a asumir un problema alternativo. Es decir la fuente la vamos a reemplazar por deltas de Dirac y la solucion por la funcion de Green $G(x,t; \xi, \tau)$ puntual en $x=\xi$, y tiempo $t=\tau$. Es decir

$$G_t = \alpha G_{xx} + \delta(x- \xi) \delta(t - \tau). \tag{3} $$

Para resolver esta funcion de Green, vamos a tomar la transforma de Fourier de $x$ a $k$

Recuerden

$$\mathcal{F}[G(x,t; \xi, \tau)] = \hat{G}(k, t; \xi, \tau) = \frac{1}{2 \pi} \int_{-\infty}^\infty G(x, t, \xi, \tau) \mathrm{e}^{-\mathrm{i} k x} dx   $$

Sacando la transformada de Fourier de (3) obtenemos

$$\hat{G_t} = -\alpha k^2 \hat{G} + \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} \delta(t - \tau) \tag{4} $$

Resolvemos la parte homogenea. Es decir estamos en un tiempo $t \ne \tau$
tenemos

$$\hat{G}_t + \alpha k^2 \hat{G} = 0 .$$
Esta ecuacion es ordinaria de primer orden. Facil

$$\hat{G}(k, t; \xi, \tau) = C(k, \xi, \tau) \mathrm{e}^{-\alpha k^2(t - \tau)}. $$

Necesitamos hallar $C(k, \xi, \tau)$

Vamos a integrar la Ecuacion (4) en un intervalo muy pequenho cercano a "la explosion" es decir cercano a $\tau$. El intervalo es
$(\tau -\epsilon, \tau+\epsilon)$.


$$\int_{\tau - \epsilon}^{\tau + \epsilon}  \frac{\partial \hat{G}(k, t; \xi, \tau)}{\partial t} dt = -\alpha k^2 \int_{\tau-\epsilon}^{\tau + \epsilon} \hat{G}(k, t; \xi, \tau) d t + \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} \delta(t - \tau) + \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi} $$

Encontramos lo siguiente

$$ \hat{G}(k, t=\tau^+; \xi) - \hat{G}(k, t=\tau; \xi) = \frac{1}{2 \pi} \mathrm{e}^{-\mathrm{i} k \xi}. $$
Por que no aparece el termino en la mitad? Vamos a tomar el limite cuando $\epsilon \to 0$. Asumimos que $\hat{G}$ es continua y por lo tanto la integral entre $-\tau, \tau$ desparece.
