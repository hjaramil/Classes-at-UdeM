## Solucion de d'Alembert:
En la clase anterior probamos la siguiente formula:

$$u(x,t) = \frac{\lambda_2 \varphi(x + \lambda_1 t) - \lambda_1 \varphi(x + \lambda_2 t)}{\lambda_2 - \lambda_1}  + \frac{1}{\lambda_2 - \lambda_1} \int_{x + \lambda_1 t}^{x + \lambda_2 t} \psi(s) ds. \tag{1} $$

A esta solucion se conoce con el nombre de **d'Alembert**.
[d'Alembert formula](https://en.wikipedia.org/wiki/D%27Alembert%27s_formula).



**Ejemplo 3.2.1**: Considere la ecuacion

$$\frac{\partial^2 u(x,t)}{\partial t^2} - c^2 \frac{\partial^2 u(x,t)}{\partial x^2}=0  \quad , \quad -\infty < x < \infty \quad , \quad t > 0   $$
con condiciones iniciales
$$
\begin{cases}
u(x,t=0) = \varphi(x) \\
u_t(x,t=0) = \psi(x) \\
\end{cases}
$$

**Solucion**: Factorizamos la ecuacion diferencial, el **polinomio caracteristico** es

$$\lambda^2 - c^2  = (\lambda - c) ( \lambda + c)$$
donde $\lambda_1 = - c$,  $\lambda_2=c$
Basicamente podemos factorizar el operador

$$u_{tt} - c^2 u_{xx} = (\partial_{tt} - c^2 \partial_{xx})u
= (\partial_t - c \partial_x )(\partial_t + c \partial_x) u$$

Usamos la Ecuacion de d'Alembert (1)
donde $\lambda_2/(\lambda_2 - \lambda_1) = 1/2$, $-\lambda_1/(\lambda_2 - \lambda_1) = -1/2$, $\lambda_2 - \lambda_1 = 2c$

$$u(x,t) = \frac{\varphi(x -ct ) + \varphi(x + ct)}{2} + \frac{1}{2c}
\int_{x-ct}^{x+ct} \psi(x) ds $$

Definamos **rango de influencia**. Dado un punto en el dominio queremos saber cual es el rango de influencia para un tiempo $t>0$.

**Ejemplo 3.2.3 **
Considere el problema  IVP

$$
\begin{cases}
u_{tt} - c^2 u_{xx} = 0  \quad , \quad -\infty < x < \infty \quad , t>0 \\
u(x,0) = \varphi(x) = \cos x \\
u_t(x,0) = \Psi(x) = \sin x \\
\end{cases}
$$

La solucion de d'Alembert es

\begin{eqnarray}
u(x,t)  &=& \frac{\varphi(x + ct) + \varphi(x - ct)}{2} + \frac{1}{2c} \int_{x - ct}^{x + ct}  \psi(s) ds    \\
&=& \frac{\cos( x + ct) + \cos(x - ct)}{2} + \frac{1}{2c} \int_{x - ct}^{x + ct} \sin s ds \\
&=& \frac{\cos(x + ct) + \cos (x - ct)}{2} - \left .   \frac{1}{2c} \cos s
\right |_{s=x-ct}^{s=x+ct} \\
&=& \frac{\cos (x + ct) + \cos(x-ct)}{2} + \frac{1}{2c} (\cos(x+ct) - \cos(x - ct))
\end{eqnarray}
Ahora si $c=1$

$$ u(x,t) = \cos(x-t)$$

Consideremos el siguiente (mas general) problema con coeficientes constantes $a,b,c,d,e,f,g$

$$a u_{xx} + b u_{xy} + c u_{yy} + d u_x + e u_y + fu + g = 0 $$

Queremos llevar esta ecuacion a la forma **canonica** (es decir eliminar los terminos puros $a,b$)
Para esto hacemos un **cambio de variables** (de $x,y$, a $\xi , \eta$) es decir

$$ u(\xi, \eta) = u(\xi(x,y), \eta(x,y)$$. En estas nuevas variables el sistema queda como


$$a^* u_{\xi \xi } + b^* u_{\xi \eta} + c^* u_{\eta \eta} + d^* u_\xi + e^* u_\eta + f^*u + g^* = 0 $$


donde (usando la regla de la cadena en derivadas parciales)

\begin{eqnarray}
a^* &=& a \xi_x^2 + 2 b \xi_x \xi_y  + c \xi_y^2 \\
b^* &=& a \xi_x \eta_x + b ( \xi_x \eta_y + \xi_y \eta_x) + c \xi_y \eta_y \\
c^* &=& a \eta_x^2 + 2 b \eta_x \eta_y + c \eta_y^2 \\
d^* &=& d \xi_x + e \xi_y + a \xi_{xx} + 2b \xi_{xy} + c \xi_{yy} \\
e^* &=& d \eta_x + e \eta_y + a \eta_{xx} + 2b \eta_{xy} + c \eta_{yy} \\
f^* &=& f \\
g^* &=& g \\
\end{eqnarray}

Observamos que si $x=\lambda_1$, $y = \lambda_2$ donde $\lambda_1, \lambda_2$ son soluciones de la ecuacion cuadratica

$$a x^2 + 2 bx + c = 0 $$
y definiendo

$$\xi = y + \lambda_1 x \quad , \quad \eta = y + \lambda_2 $$
entonces $a^*=b^*=0$ (tambien va en la proxima tarea).
y por lo tanto

$$ b^* u_{\xi \eta} + d^* u_\xi + e^* u_\eta + f^*u + g^* = 0 $$


**Ejemplo 3.2.4**: Encuentre la solucion general de

$$2 u_{\xi \eta} - u_\eta = 0  $$

Llamamos $v=u_{\eta}$, entonces tenemos

$$2 v_{\xi} - v = 0  $$
Integramos

$$2 \frac{dv}{d \xi} = 0 \implies  2 \int \frac{dv}{v} + C(\eta) = d \xi $$
Es decir

$$2 \ln v + C(\eta) = \xi  $$
de donde tomando exponencial



$$v(\xi, \eta) = \varphi(\eta) \mathrm{e}^{\xi/2} $$
donde $\varphi(\eta) = \mathrm{e}^{-C(\eta)/2}$ .

Tenemos  que integrar una vez mas, (recuerde que $v=u_\eta$) es decir,

$$\frac{du}{d \eta} = \varphi(\eta) \mathrm{e}^{\xi/2} \implies  du = \varphi(\eta) \mathrm{e}^{\xi/2} d \eta $$
e integrando con respecto a $\eta$

$$u = \int \varphi(\eta) \mathrm{e}^{\xi/2} d \eta =\mathrm{e}^{\xi/2} \int \varphi(\eta) d \eta + \psi(\xi).  $$

Podemos hacer un renombramiento

$$ F(\eta) = \int \varphi(\eta) d \eta $$
$$ G(\xi) = \varphi(\xi)$$
con esta nueva notacion tenemos que la solucion general es


$$u = \mathrm{e}^{\xi /2} F(\eta) + G(\xi). $$

## Coeficientes variables.
Considere la ecuacion (**ojo esta vez en $x,t$)

$$Lu = a(x,t) u_{tt} + 2 b(x,t) u_{tx} + c(x,t) u_{xx} + d(x,t) u_t + c(x,t) u_x + f(x,t) u = g(x,t) \quad , \quad x \in \mathbb{R} \quad , t > 0 \tag{1a} $$

Por ele momento ignoramos las condiciones iniciales. Asumimos
que en un punto $(x,t)=(x_0,t_0)$

$$D = \left .  b^2(x,t) - a(x,t) c(x,t) \right |_{x=x_0, t=t_0}  > 0 $$

### Cambio de coordenadas.
Queremos hallar unas nuevas coordenadas $\xi, \eta$ tales que en las nuevas coordenadas los terminos puros (en $a,c$) se cancelen (sean 0).

Si esto sucediera entonces para $u=u(x(\xi, \eta), y(\xi, \eta))$

$$Lu = 2 b^*(x,t) u_{tx}  + d^*(x,t) u_t + c^*(x,t) u_x + f^*(x,t) u = g^*(x,t) \quad , \quad x \in \mathbb{R} \quad , t > 0 $$


la transormacion en las nuevas variables es

$$\xi = \xi(x, t) $$
$$\eta = \eta(x, t) $$

Queremos el jacobiano no sea 0.

$$J = \begin{vmatrix} \xi_t & \xi_x \\ \eta_t & \eta_x  \end{vmatrix} = \xi_t \eta_x - \xi_x \eta_t $$

Vamos a usar la regla de la cadena.

* Primeras derivadas con respecto a $t, x$.

$$ u_t = u_\xi \xi_t + u_\eta \eta_t \tag{2}  $$
$$ u_x = u_\xi \xi_x + u_\eta \eta_x \tag{3}  $$

* Segunda derivada $tx$.

\begin{eqnarray}
u_{tx} &=& (u_t)_x = (u_\xi \xi_t + u_\eta \eta_t) \\
&=& (u_\xi \xi_t)_x + (u_\eta \eta_t)_x \\
&=& (u_\xi)_x \xi_t + u_\xi \xi_{tx} + (u_\eta)_x \eta_t + u_\eta \eta_{tx}
\end{eqnarray}
Es decir
$$ u_{tx} = (u_{\xi \xi} \xi_x + u_{\eta \xi} \eta_x ) \xi_t + u_\xi \xi_{tx} + (u_{\eta \xi} \xi_x + u_{\eta \eta} \eta_x) \eta_t + u_\eta \eta_{tx} \tag{4} $$

De la misma forma

$$ u_{tt} = (u_{\xi \xi} \xi_t + u_{\eta \xi} \eta_t ) \xi_t + u_\xi \xi_{tt} + (u_{\eta \xi} \xi_t + u_{\eta \eta} \eta_t) \eta_t + u_\eta \eta_{tt} \tag{5} $$

$$ u_{tt} = u_{\xi \xi} \xi_t^2 + 2 u_{\xi \eta} \xi_t \eta_t + u_{\eta \eta} \eta_t^2 + u_\xi \xi_{tt} + u_\eta \eta_{tt}  \tag{6} $$



y cambiando los indices $t$ por $x$

$$ u_{xx} = u_{\xi \xi} \xi_x^2 + 2 u_{\xi \eta} \xi_x \eta_x + u_{\eta \eta} \eta_x^2 + u_\xi \xi_{xx} + u_\eta \eta_{xx}  $$

Insertamos (2),(3),(4),(5),(6) en la (1a) y agrupamos terminos, obtenemos


$$a^* u_{\xi \xi} + 2 b^* u_{\eta \xi} + c^* u_{\eta \eta} + d^* u_\xi + e^* u_\eta + f^* u = g^* $$

Donde

\begin{eqnarray}
a^* &=& a \xi_t^2 + 2 b  \xi_t \xi_x + c \xi_x^2 \\
b^* &=& a \xi_t \eta_t + b(\xi_t \eta_x + \xi_x \eta_t ) + c \xi_x \eta_x \\
c^* &=& a \eta_t^2 + 2 b \eta_t \eta_x  + c \eta_x^2 \\
d^* &=& a \xi_{tt} + b \xi_{tx} + c \xi_{xx} + d \xi_t + e \xi_x \\
e^* &=& a \eta_{tt} + b \eta_{tx} + c \eta_{xx} + d \eta_t + e \eta_x \\
f^* &=& f \\
g^* &=& g \\
\end{eqnarray}
