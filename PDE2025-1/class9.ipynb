# Ecuaciones en derivadas parciales hiperbolicas.
## Coficientes constantes.

Consideremos la siguiente ecuacion

$$L u  \equiv a \frac{\partial^2 u}{\partial t^2} + 2 b \frac{\partial^2 u}{\partial t \partial x} + c \frac{\partial^2 u}{\partial x^2}  = 0 \quad , \quad x \in \mathbb{R} \quad , t > 0  \quad , \quad a,b,c \in \mathbb{R} \tag{1} $$
A esto se le llama un IVP: Initial Value Problem. Para la solucion $u=u(x,t)$
con la condiciones iniciales

$$ \begin{cases}
u(x,0) = \varphi(x)  \tag{a} \\
\left .  \frac{\partial u(x,t)}{\partial t} \right |_{t=0} = \psi(x).
\end{cases} $$

* La primera condicion establece la forma de la cuerda (o la barra).
* La segunda nos habla de la "particle velocity" velocidad de vibracion en cada punto de la cuerda uni-dimensional.

Vamos a reducir el sistema sistema (1). Para esto pensamos algebraicamente en el metodo de completacion de cuadrados para remover el termino cruzado.

Llamemos

$$T = \frac{\partial }{\partial t} \quad , \quad   
X = \frac{\partial }{\partial x}  $$
La ecucion (1) se puede escribir como

$$(a T^2 + b T X + c X^2) u = 0 $$
Ovidemonos del $u$ por ahora

$$(a T^2 + b T X + c X^2)  = 0 $$

Hallamos las raices cuadraticas para factorizar esta expresion


$$\frac{-2 b X \pm \sqrt{ 4 b^2 X^2 - 4 ac X^2}}{2a} = \frac{-b \pm \sqrt{b^2 - ac}}{a} X  $$

Las dos raices (ignorando $X$

$$\lambda_{1,2} = \lambda_{\pm} =\frac{-b \pm \sqrt{b^2 - ac}}{a} $$
Asumimos que $b^2 - ac > 0$ (son hiperbolicas) de otra forma la raiz seria imagnaria (complejos).

Podemos entonces factorizar la ecuacion (1) como

$$
\left ( \frac{\partial}{\partial t} - \lambda_1 \frac{\partial}{\partial x}  \right )
\left ( \frac{\partial}{\partial t} - \lambda_2 \frac{\partial}{\partial x}  \right ) u = 0.
 $$
 De aca que si $UV=0$ entonces $U=0$ O $V=0$, de forma que tenemos las dos ecuaciones de primer orden

 $$
 \frac{\partial u}{\partial t} - \lambda_1 \frac{\partial u}{\partial x} = 0
 \quad , \quad  
 \frac{\partial u}{\partial t} - \lambda_2 \frac{\partial u}{\partial x} = 0  
 $$

En este momento recordamos la teoria del primer capitulo (las caracteristicas)


$$ \frac{dx}{dt} = - \lambda_1  \quad , \quad \frac{dx}{dt} = - \lambda_2 $$
con $\lambda_{1,2}$ reales. Las soluciones son (interando)

$$
x = -\lambda_1 t + k_1  \quad , \quad  x = -\lambda_2 t + k_2
$$

La Figura siguiente muestra las caracteristicas para $\lambda_1=-1, \lambda_2=1$

Si definimos dos nuevas variables o (despejando las constantes $k_1, k_2$ de arriba.

$$\xi = x + \lambda_1 t  \quad , \quad \eta = x + \lambda_2 t $$
entonces vemos que la solucion es

$$u(x,y) = f(\xi) + g(\eta).  $$
La iterpretacion es de dos ondas viajando en direcciones contrarias (por que asumimos que $\lambda_1 \lambda_2 < 0 $
$f ( x + \lambda_1 t) $ onda viajando a la velocidad $-\lambda_1$ en la direccion $x$
$f ( x + \lambda_2 t) $ onda viajando a la velocidad $-\lambda_2$ en la direccion $x$

Se puede probar (se deja de tarea) que la ecuacion diferencial (1)  se reduce a

$$ u_{\xi \eta} = \frac{\partial^2 u}{\partial \xi \partial \eta} = 0 \tag{2} $$
A esta ecuacion se le conoce con el nombre de **forma canonica de la PDE de segundo orden**
Resolvamos la ecuacion (2).

Asuamos que

$$\frac{\partial u}{\partial \eta}(\xi, \eta) = v(\xi, \eta) \tag{3} $$
Esta es una variable auxiliar.

Entonces, de la ecuacion (2)
$$ \frac{\partial v}{\partial \xi} = \left . \frac{d v}{d \xi} \right |_{\eta=\text{const}} = 0 $$
De forma que
$$v = \text{const}$$ para $\eta=\text{const}$, es decir $v$ no depende de $\xi$.
De forma que podemos escribir

$$v(\xi, \eta) = g( \eta).  $$

De forma que de la ecuacion (3)

$$ \left . \frac{d}{d \eta} u \right |_{\xi = \text{const}} = g(\eta)  $$

Integramos esta ecuacion con respecto a $\eta$

$$ \left . u \right |_{\xi= \text{const}} = \int g(\eta) d \eta) + f(\xi). $$

De forma que

$$u(x,t) = f(\xi) + g( \eta) = f(x + \lambda_1 t) + g(x + \lambda_2 t)  \tag{4}$$
donde $f,g$ son funciones de una variable y redefinimos

$$g(\eta) \leftarrow \int g(\eta) d \eta $$

## Datos iniciales (IVP)
Para encontrar $f,g$ usamos las condiciones iniciales (a) (desplazamiento y velocidad de particula) usando la ecuacion (4).

desplazamiento $$u(x,0) = f(x) + g(x) = \varphi(x)  \tag{5} $$

velocidad de particula $$ \frac{\partial u(x,0)}{\partial t} = \lambda_1 f'(x) + \lambda_2 g'(x)= \psi(x)  \tag{6} $$

Las dos ecuaciones (5) y (6) estan en las indefinidas $f,g$. Debemos
resolver estas dos ecuaciones con las dos incognitas $f,g$.

Integremos esta segunda ecuacion entre 0 y $x$

$$\lambda_1 f(x) + \lambda_2 g(x) = \int_0^x \psi(s) ds \tag{6} $$

Despejemos $f$ de (5)

$$ f(x)= \varphi(x) - g(x) \tag{7} $$
insertamos esta en la

$$\lambda_1 (\varphi(x) - g(x) )+ \lambda_2 g(x) = \int_0^x \psi(s) ds  $$

Despejemos $g(x)$ de aca

$$g(x) = \frac{1}{\lambda_2 - \lambda_1} \int_0^x \psi(s) ds - \frac{\lambda_1}{\lambda_2 - \lambda_1} \varphi(x) \tag{8}  $$

Pregunta para la tarea: Por que puedo dividir por $\lambda_2 - \lambda_1$, sera que $\lambda_1 \ne \lambda_2$? y por que?

Ya la ecuacion (7) nos da la solucion de $g$ en terminos de el desplazamiento y la velocidad de particula. Falta $f$, comenzamos con (7) e insertamos $g$


\begin{eqnarray}
f(x) &=& \varphi(x) - g(x)  \\
&=&  \left (  1 + \frac{\lambda_1}{\lambda_2 - \lambda_1} \right ) \varphi(x)   - \frac{1}{\lambda_2 - \lambda_1} \int_0^x \psi(s) dx \\
&=& \frac{\lambda_2}{\lambda_2 - \lambda_1} \varphi(x) - \frac{1}{\lambda_2 - \lambda_1} \int_0^x \psi(s) ds.
\end{eqnarray}

Ya hallamos $f,g$ en terminos de los datos iniciales (desplazamiento en $t=0$, $\varphi(x)$ y velocidad de particula en $t=0$ $\psi(x)$.
De forma que

$$u(x,t) = \frac{\lambda_2 \varphi(x + \lambda_1 t) - \lambda_1 \varphi(x + \lambda_2 t)}{\lambda_2 - \lambda_1}  + \frac{1}{\lambda_2 - \lambda_1} \int_{x + \lambda_1 t}^{x + \lambda_2 t} \psi(s) ds. $$

A esta solucion se conoce con el nombre de **d'Alembert**.
[d'Alembert formula](https://en.wikipedia.org/wiki/D%27Alembert%27s_formula).


**Definicion**: Dominio y rango de dependencia.
La solucion en punto $(x_0, t_0)$ depende del intervalo $(x_0 + \lambda_1 t, x_0 + \lambda_2 t)$. Veamos como, de forma grafica.
