**Ejemplo 2.2.1**: Este ejemplo tomado de Bleistein
Considere

$$ u_x + u_y = u  $$
$a=1, b=1, c=u$

Las caracteristica son

$$ \frac{dx}{1} = \frac{dy}{1} = \frac{du}{u} $$

El algoritmo es:

1. De las dos primeras encontrar $y = y(x)$
2. De la primera y la tercera encontrar $u=u(x)$.


Paso 1. Dividamos las dos primeras

$$ \frac{dy}{dx} = 1 $$
Solucion $y= x + c_1$.
Dividamos la tercera por la primera

$$ \frac{du}{dx} = u  $$
Solucion
$$ u(x) = c_2 \mathrm{e}^x \tag{6}$$
Es decir tenemos

$$y = x + c_1 \quad , \quad u = c_2 \mathrm{e}^x $$
Todavia no sabemos quien es $c_1, c_2$.

Podemos pensar que amarramos $c_2$ a $c_1$ mediante una funcion $f$. Por ejemplo

$$c_2 = f(c_1) $$

Es decir, de 6 $c_2 = \mathrm{e}^{-x} u$

$$\mathrm{e}^{-x} u = f(c_1) = f(y-x) $$


$$ u = \mathrm{e}^{x}  f(y-x) $$

Pero aun asi no sabemos quien es $f$? $f$ puede ser muchas cosas.
Toca definir $f$, para $x=0$, $u=f(y)$

Por ejemplo si $u=y^2$, $x=0$ (condicion inicial)
Ya con esto la solucion es

$$ u = \mathrm{e}^x ( y- x)^2 $$
A esta solucion se le conoce (Bleistein) como **envelope condition** (mas adelante estudiamos el concepto de "envelope" (envolvente)

 $$

 Los coeficientes que vamos a escoger son $a=b=1, c= u$

 $$x = \sigma + x_0(\tau) \quad , \quad y = \sigma + y_0(\tau) \quad , \quad u=u(\tau) \mathrm{e}^{\sigma}$$

 De forma que si las condiciones iniciales dan

 $$ x_0(\tau) = \tau  \quad , \quad y_0(\tau) = - \tau \quad, \quad u_0(\tau) = \tau $$

\begin{cases}

**Ejemplo 2.2.2 (el caso de $J=0$ a lo largo de la curva inicial):
Asumamos que la curva inicial esta dada por

$$x_0(\tau) = y_0(\tau) = u_0(\tau) = \tau $$

En terminos de $\sigma, \tau$

$$x = \sigma + \tau \quad , \quad y = \sigma + \tau \quad , \quad u = \tau \mathrm{e}^{\sigma} $$

El Jacobiano de este sistem es:

$$J =
\begin{vmatrix}
x_\sigma & x_\tau \\
y_\sigma & y_\tau
\end{vmatrix} =
\begin{vmatrix}
1 & 1 \\ 1 &  1
\end{vmatrix} = 0 $$
a lo largo de toda la curva inicial. La curva inicial es $x=y$

Recuerden del "envelope condicion"
$$u = \mathbb{e}^x f(y-x) $$
Como $x=y$ entonces
$$u = \mathbb{e}^x f(0) $$

Esta solucion no es unica por que $f$ es arbitrario, pero eso no tiene nada que ver con que $J=0$. Pero si tiene que ver con que $J=0$ que se perdio la $y$,
la solucion es una fucion de una sola variable $u=u(x)$. Esta solucion es **degenerada**. No es una solucion, donde estan los $y$?

**Ejemplo 2.2.3** (el caso de el jacobiano $j=0$ en un punto)
Asumamos la curva inicial (datos)

$$ x_0(\tau) = \tau  \quad , \quad y_0(\tau) = \frac12 \tau^2 $$

De forma que
$$x = \sigma + \tau \quad , \quad y = \sigma + \frac12 \tau^2 $$

El Jacobiano de esto es
$$
\begin{vmatrix}
x_\sigma & x_\tau \\
y_\sigma & y_\tau
\end{vmatrix} =
\begin{vmatrix}
1 & 1 \\ 1 &  \tau
\end{vmatrix} = \tau - 1 $$

Este  Jacobiano es 0 solo si $\tau=1$. Es decir en un punto.

Vamos a hallar la solucion $u=u(x,y)$.
Debemos invertir el sistema

$$
\begin{cases}
x = \sigma + \tau  \\
y = \sigma + \frac12 \tau^2
\end{cases}
$$


restando

$$x - y = \tau - \frac12 \tau^2 $$
Agrupando
$$\frac12 \tau^2 - \tau + (x - y) = 0 $$

Resolvemos la cuadratica por $\tau$

$$\tau = \frac{1 \pm \sqrt{1 - 2(x-y)}}{2(1/2} $$
$$\tau =  \pm \sqrt{1 - 2(x-y)}$$

Dos soluciones. Recuerden que $x= \sigma + \tau$, entonces

$$\sigma = x - \tau $$
$$\sigma = x  \mp \sqrt{1 - 2(x-y)} $$
De forma que la solucion es

$$u(x,y) = \left (  1 \pm \sqrt{1 - 2(x-y)} \right )
\mathrm{e}^{x \mp \sqrt{1 - 2(x-y)}}.$$

Donde esta el problema de que $J = 0$? esta en que si vamos a calcular las derivadas $u_x, u_y$ en el denominador aparece
la raiz cuadrada

$$\sqrt{1 - 2(x-y)}$$
Esta raiz cuadrada es 0 si $2(x-y)=1$. Y eso es problema (division por cero).

Analicemos la situacion

* Las caracteristica estan dadas por lines de 45 gradoes $y = x + c$.

* La singularidad esa cuando $1 - 2(x-y)=0$, es decir

$$ y = x - 1/2 $$

* La proyeccion de los datos iniciales en el plano $xy$ producen la curva parametrica

$$x = \tau \quad , y=\tau^2/2 $$

Esta curva inicial es una parabola.  Ahora que pasa en el punto $y=x-1/2$?
para (1,1/2)$

$$\left . \frac{dy}{dx} \right |_{x = 1} = 1  \text{para la linea } y = x - 1/2 $$
$$\left . \frac{dy}{dx} \right |_{x = 1} = \frac12 (2) x = x = 1 \quad , \quad \text{para}  y = x^2/2  $$

En resumen vimos dos casos:

1. $J=0$ a lo largo de toda la curva inicial (datos). Solucion degenerada.
2. $J=0$ en un punto. El problema es que en ese punto no existe la derivada de $u$, ni $u_x$ ni $u_y$. (division por cero).

**Ejemplo 2.2.4** (caracteristicas que se cruzan): Considere el siguiente IVP (Initial Value Problem)

 $$\begin{cases} x u_x + y u_y = \alpha u  \\ u(x,1)=u_0(x) \end{cases}  $$

 Claramente la condicion inicial indica que $y=1$.

image.png

Del sistema de arriba tenemos que $a=x, b=y, c= \alpha u$. De las caracteristicas tenemos

$$ \frac{dx}{d\sigma} = x  $$
$$ \frac{dy}{d\sigma} = y  $$

De aca (eliminamos los parametros, dividiendo la 2 por la 1)

$$ \frac{dy}{dx} = \frac{y}{x} $$
Resolvamos esta ecuacion difierencial ordinaria (separacion de variables)

$$ \frac{dy}{y} = \frac{dx}{x} $$
$$ \log y = \log x + c_1 $$ (integrando)

Sacamos la exponencial

$$ y = \mathrm{e}^{c_1} x $$
Podemos reescribir esto como
$$ y = c x $$

Es decir, las caracteristicas son de la forma $y=cx$. Lineas que pasan por el origen y que tienen pendiente $c>0$.

Veamos la grafica.

image.png

En la grafica se observa que los datos iniciales $y=1$ cruzan las caracteristicas (eso esta bien). Pero que pasa en el punto $(0,0)$.

Veamos la solucion $u(x)$ a largo de una caracteristica $y=cx$, $c>0$.
Es decir $u(x) = u(x,y) = u(x, cx)$

Usamos la regla de la cadena para derivar

$$\frac{du}{dx} = u_x + c u_y = u_x + \frac{y}{x} u_y = \frac{x u_x + y u_y}{x} = \frac{\alpha u}{x} $$
donde usamos que $c=y/x$.  Integramos esta ecuacion usando separacion de variables.

$$ \frac{du}{u} = \frac{\alpha dx}{x} $$
$$ \log u = \alpha \log x + c_2 $$,
exponenciando
$$ u = \mathrm{e}^{c_2} \mathrm{e}^{\alpha \log x} $$
llamemos $k= \mathrm{e}^{c_2}$

$$ u = k \left ( \mathrm{e}^{\log x} \right)^{\alpha}$$
$$ u = k x^{\alpha}$$.

Econtramos enconces que

$$ u(x, cx) = k x^{\alpha}$$

$$ u(x, cx) = k \left ( \frac{y}{x}  \right ) x^{\alpha} \tag{1} $$

Vemos que $k$ es funcion de $c=y/x$, podemos
escribir

$$u_0(x) = k \left ( \frac{y}{x} \right )  x^{\alpha} $$
Debemos hallar $k$.

$$u_0 \left ( \frac{y}{x} \right ) = k \left ( \frac{x}{y} \right )  \frac{1}{x^{\alpha}} $$

De aca que
si $y=1$

$$u_0 \left ( \frac1x \right ) = k(x) \frac{1}{x^{\alpha}}   $$


Despejamos $k(x)$.

$$k(x) = u_0 \left ( \frac{1}{x}  \right ) x^{\alpha}  $$

$$k \left ( \frac{y}{x} \right ) = u_0 \left (  \frac{x}{y}  \right )
\frac{y^\alpha}{x^\alpha}   $$

De forma que

$$u(x,y) = u_0 \left (  \frac{x}{y} \right ) y^{\alpha} $$
de la ecuacion (1)

Es suficiente proveer datos lejos de $(0,0)$ para encontrar la solucion.


## Ecuaciones quasilineales de primero orden (PDEs)
La ecuacion que queremos resolver es:

$$ a(x,y,u) u_x + b(x,y,u) u_y = c(x,y,u)$$

Si seguimos los mismos pasos que hicimos cuando resolvemos la ecuacion lineal obtenemos las mismas caracteristicas pero con coeficientes dependiendo de $u$.

Es decir, recuerden que se crearon dos vectores , uno tangenge $t$ y uno con con los coeficientes $v = [a(x,y,u), b(x,y,u), c(x,y,u)]$ que son perpendicuales al gradiente y por lo tanto en el plano tangente. Se escoje la curva  caracteristica a lo largo de la tangente.  Tenemos por definicion de paralelismo en algebra lineal $t = \lambda v$. Es decir

$$ \frac{dx}{d \sigma} = \lambda a(x,y, u)  
\quad , \quad  \frac{dy}{d \sigma} = \lambda b(x,y, u) \quad , \quad
\frac{du }{d \sigma} = \lambda c(x,y, u)  $$


Algunas veces se escribe de forma simplificada

$$\frac{dx}{a} = \frac{dy}{b} = \frac{du}{c} $$
y si la solucion se puede integrar, dividiendo (para evitar parametrizacion)

$$\frac{dy}{dx} = \frac{b(x,y,u)}{a(x,y,u)} \quad , \frac{du}{dx} = \frac{c(x,y,u)}{a(x,y,u)}. $$
