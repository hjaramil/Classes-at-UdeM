Abandonamos la variable $t$ (tiempo) y agregamos la variable espacial $y$.

La ecucacion de **Laplace** es:

$$\nabla^2 u(x,y) = 0 $$
donde

$$\nabla^2 \equiv \nabla \cdot \nabla = \frac{\partial^2}{\partial x^2} + \frac{\partial^2}{\partial y} $$

Considermos condiciones de bordo en una caja $[0,a] \times [0, b]$.

Las siguientes condiciones de frontera.

$$
\begin{cases}
u(0, y) = u(a,y) = u(x,0) = 0  \\
u(x,b) = \varphi(x).
\end{cases}
$$

### Separacion de variables

$$u(x,y) = F(x) G(y) $$
De aca

$$ F''(x) G(y ) + F(x) G''(y) = 0 $$
Dividiendo por $F(x) G(y)$,

$$\frac{F''(x)}{F(x)} = -\frac{G''(y)}{G(y)} = k $$

$k$ es constante por los argumentos ya explicados varias veces.
Se originan dos ODE (Ordinary Differential Equations)

$$F''(x) = k F(x) \quad, \quad G''(y) = -k G(y) $$

Como siempre vamos a considerar tres valores de $k$

1. $k=0$: De $F''(x)=0$ tenemos que

$$u(x,y) = F(x) G(y) = (C_1 + C_2 x) G(y) $$
Aplicamos condiciones de bordo

$$ u(0,  y) = 0 \implies C_1 G(y) =0  $$
Como no queremos $G(y)=0$ entonces $C_1=0$.
En el bordo derecho

$$ u(a,  y) = 0 \implies C_2 a G(y) =0  $$
con el mismo argumento vemos que $C_2=0$. De forma que $F(x)=0$ y la solucion es
$u(x,t)=0$. No nos interesa.

2. $k>0$.

El resto del problema se resolvio en el tablero (se fue la luz). Lo copio y pego de la clase correspondiente del semestre anterior.

2. $k>0$, es decir $k = \lambda^2$, $\lambda \ne 0$.
Tenemos la ecuacion para $F$. La ecuacion

$$F'' - \lambda^2 F=0 $$
Pensemos el operador $D=\partial / \partial x$ entonces

$$(D - \lambda)(D + \lambda) F = 0 $$
Las soluciones son $\mathrm{e}^{\lambda x} $ y $\mathrm{e}^{-\lambda x}$
la general

$$F(x) = C_1 \mathrm{e}^{\lambda x} + C_2 \mathrm{e}^{-\lambda x} $$
La condicion $u(0,y)$ indica
$$u(0,y) = F(0) G(y) = 0$$
$$F(0)=0 \implies C_1 + C_2 = 0 $$
o sea que $C_1 = - C_2$, y

$$F(x) = C_1 ( \mathrm{e}^{\lambda x} - \mathrm{e}^{-\lambda x}) =2 C_1 \sinh \lambda x $$

Esto es cero solo si $\lambda=0$ pero ya vimos que esto produce la solucion trivial y $C_1=0$, $C_1=-C_2=0$, la trivial

3. $k < 0 $, en cuyo caso $k=-\lambda^2$.
La ecuacion diferencial para $F$ es

$$F'' + \lambda^2 F = 0 $$
Las soluciones son  $\mathrm{e}^{\mathrm{i} \lambda x}$ o
$\mathrm{e}^{-\mathrm{i} \lambda x}$
o en terminos de senos y cosenos,

$$F(x) = C_1 \sin \lambda x + C_2 \cos \lambda x  $$
de $F(0)=0$ entonces $C_2=0$ y

$$ F(x) = C_1 \sin \lambda x $$
De la segunda condicion $F(a)=0$,

$$F(a) = C_1 \sin \lambda a = 0 $$
Como queremos que $C_1 \ne 0$, entonces

$\lambda a = n \pi$.
De forma que

$$\lambda_n = \frac{n \pi}{a} $$
$$ F(x) = C_1 \sin \left ( \frac{n \pi x}{a}  \right ) .$$

Resolvamos $G# la ecuacion es

$$G''(y) - \lambda G(y) = 0.  $$
o sea

$$G''(y) = \left ( \frac{ n \pi}{a} \right)^2 G(y) = 0 $$


La solucion en terminos de exponenciales es:

$$G_n(y) =  C_3 \mathrm{e}^{\frac{n \pi}{a} y}   + C_3 \mathrm{e}^{-\frac{n \pi}{a} y} $$

Condiciones iniciales. En $u(x,0)=F(x)G(y)$, $G(0)=0$

$$ C_3 + C_4 = 0$$
o sea que que $C_3 = - C_4$,

$$G_n(y) =  C_3 ( \mathrm{e}^{\frac{n \pi}{a} y}  -  \mathrm{e}^{-\frac{n \pi}{a} y}) =  2 C_3  \sinh \frac{n \pi}{a} y$$

Vamos a $u$ y por separacion de variables

$$u(x,y) = F(x) G(y) = 4 C_1 C_3 \sin \left ( \frac{n \pi}{a} x  \right )\sinh \left ( \frac{n \pi}{a} y \right )$$

El coeficiente $4 C_{1n} C_{3m}$ lo llamamos $a_{n}$ aplicamos superposicion y encontramos

$$u(x,y) = \sum_{n=1}^{\infty} a_n \sin \frac{n \pi x}{a} \sinh \frac{n \pi y}{a} \tag{1} $$

Recuerden la condicion de borde $u(x,b)=\varphi(x)$.

$$u(x,b) = \sum_{n=1}^{\infty} a_n \sin \frac{n \pi x}{a} \sinh \frac{n \pi b}{a} $$

Toca hallar los $a_n$. Para esto proyectamos la condicion inicial $\varphi(x)$ sobre las autofunciones

$$f_n(x) = \sin \frac{n \pi x}{a} $$
que son ortogonales. Encontramos de clase anterior

$$\langle f_i, f_j \rangle = \frac{a}{2} \delta_{ij}   $$

Entonces, la proyeccion de la condicion inicial $\varphi(x)$ sobre la autofuncion $\sin ( j \pi x/a)$


\begin{eqnarray}
\left \langle \varphi(x) , \sin \frac{ j \pi x}{a}  \right \rangle &=& \sum_{n=1}^{\infty} a_n \left \langle \sin \frac{n \pi x}{a}, \sin \frac{j \pi x}{a} \right \rangle \sinh \frac{n \pi b}{a} \\
&=& \sum_{n=1}^{\infty} \frac{a}{2} \delta_{nj} a_n \sinh \frac{n \pi b}{a} \\
&=& \frac{a}{2}  a_j \sinh \frac{j \pi b}{a} \\
\end{eqnarray}
De donde, despejando $a_j$,



$$a_j =  \frac{2}{ a \sinh (j \pi b/a)} \left \langle \varphi(x), \sin \frac{j \pi x}{a} \right  \rangle =\frac{2}{ a \sinh (j \pi b/a)}
\int_0^a \varphi(x) \sin \frac{ j \pi x}{a} dx \tag{2}  $$

La solucion esta dada por (1) con $a_j$ en (2)

Es decir:

$$u(x,y) = \begin{cases}
u(x,y) = \sum_{n=1}^{\infty} a_n \sin \frac{n \pi x}{a} \sinh \frac{n \pi y}{a}  \\
a_n= \frac{2}{ a \sinh (n \pi b/a)} \sin_0^a \varphi(x) \sin \frac{n \pi x}{a} dx
\end{cases}
$$
# Ecuaciones lineales PDE con fuente.
Estas se pueden escribir como

$$L u = f \tag{2}$$
donde $L$ es un operador lineal (por ejemplo el SLO, ecuacion de onda o de calor, o de Laplace). En este momento resolvimos el problema $Lu=0$. Con condiciones iniciales y de frontera que lo que hacen es encontrar una solucion unica. De otra forma veamos que la solucion no seria unica.
Definamos el espacio nulo (como en algebra lineal)

$$\mathcal{N}(L) = \{ u | Lu = 0 \} $$
no imponenmos ni condiciones de contorno ni iniciales.
Entonces si $u_p$ es una solucion de (2) $Lu_p = f$.
Entonces cualquier solucion de la forma

$$u = u_p + u_N $$
donde $u_N \in \mathcal{N}(L)$, es $L u_N = 0$ tambien es solucion de (2)

Pues

$$Lu = L(u_p + u_N) = L(u_p) + L(u_N) =  f + 0 = f$$
o sea que $u$ es solucion de (2). Hay infinitas si el espacio nulo no es el trivial $\{ 0 \}$.

Sin embargo nosotros no hemos ensenhado a resolve el problema $Lu=f$ con $f \ne 0$.

Para poder aborar este capitulo necesitamos introducir un concepto nuevo que es **teoria de distribuciones**. La idea detras de la teoria de distribuciones es que son como una generalizacion de funciones, por esto tambien se les llama **funciones generalizadas**. Estas tienen el beneficio de de tener todas derivadas que se quieran.  Vamos a ver por que lo que digo es cierto.
Tambien se pueden integrar cosas que normalemente no son integrables.

Antes de definir las funciones generalizadas (distribuciones) definamos las funciones de prueba **test functions**:

### Funciones de prueba
El espacio de funciones de prueba $\mathcal{D}$, consiste en todas las funciones que son inifinito-diferenciales (que tienen un numero infinito de derivas) en un dominio cerrado y acotado. En los reales es un intevalor $[a,b]$ o una union finita de estos.

$$\mathcal{D} = \{ \varphi \in C^{\infty}(\mathcal{R}) | \varphi \text{   tiene soporte compacto}   \} $$
compacto en los reales quiere decir cerrado y acotado.


El espacio $\mathcal{D}$ no es vacio. Veamos un ejemplo

$$\varphi(x) =
\begin{cases}
\mathrm{e}^{-\frac{1}{1 - x^2}}  \quad ,  \quad \text{si } |x | < 1 \\
0 \quad , \quad \text{si } |x | \ge 1.
\end{cases} $$

Veamos esta funcion en `Python`

### Distribuciones (funciones generalizadas)
Una distribucion es un operador lineal $T$ de la forma

\begin{eqnarray}
T: \mathcal{D}  &\to& \mathbb{R} \\
\varphi &\mapsto& T(\varphi)
\end{eqnarray}

Las distribuciones son lineales por definicion. Es deciro

$$T( a \varphi_1 + b \varphi_2)= a T(\varphi_1) + b T(\varphi_2) $$
donde $a, b \in \mathbb{R}$ y $\varphi_1, \varphi_2 \in \mathcal{D}$.


Vamos a usar el **producto interno** como una maquina de generar distribuciones.

Es decir

$$T(\phi) = \langle T(x), \phi(x) \rangle $$
Si la notacion parece ambigua (y lo es) piense en

$$T_f(\phi) = \langle f(x), \phi(x) \rangle $$

#### Ejemplos de distribuciones
##### El Dirac Delta.
Un delta de Dirac (Paul Dirac) define una distribucion la cual toma una funcion $\varphi$ y la evalua en 0.

$$\delta( \varphi(x)) = \varphi(0) $$
Para cualquier funcion de prueba $\varphi(x)$ se escribe comunmente

$$\langle \delta(x), \varphi(x) \rangle = \varphi(0) $$
Es decir, usando la notacion integral

$$\int_{-\infty}^\infty \delta(x) \varphi(x) dx = \varphi(0) $$

Esta ultima propiedad se llama el **sifting property of the Delta**. Se puede generalizar de la siguiente forma

$$\langle \delta(x-x_0), \varphi(x) \rangle = \varphi(x_0). $$
Dejo como tarea probar esto (cambio de variable $u=x-x_0$.

Una forma de tener un concepto intuitivo de lo que es un Delta de Dirac es a traves de secuencias de funciones. Es decir, existe una secuencia de funciones $f_n= \delta_n$ , $n \in \mathbb{N}$ tal que

1. $\delta_n(x) \ge 0$, $\forall x \in \mathbb{R}$
2. $\int_{-\infty}^{\infty} \delta_n(x) dx = 1$

Veamos dos ejemplos. Una familia de rectangulos todos con area 1 centrados en el origen

\begin{cases}

$$
