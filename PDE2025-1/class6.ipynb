# Algoritmo general para resolver PDEs no lineales de primer orden.
La ecuacion general se puede escribir como

$$F(x, y, u, u_x, u_y) = 0. $$
Es comun llamar $u_x=p$, $u_y=q$.

$$F(x, y, u, p, q) = 0. $$
para simplificar notacion o escritura.

Tres puntos intersantes

* La geometrica de estas ecuaciones no es tan simple como las quasi-lineales. Aca en vez de curvas caracteristicas veremos cintas caracteristicas. En vez de caracteristicas tangentes veremos superficies tangentes llamadas **Monge Cones**.


* Es mas probable encontrar problemas de
    * Existencia: Por ejemplo la ecuacion $p^2 + q^2 + 1= 0$, no tiene solucion.
    * Unicidad. Esto lo vemos mas adelante.

* El problema IVP es similar. Mediante datos iniciales sobre una curva se halla la superficie integral $u(x,y)$.

Comenzamos. Vamos a discutir varios aspectos:
    * Monge cones
    * Cintas caracteristicas (characteristic strips)
    * Ecuaciones characteristicas (en este caso son 5)

Por el momento vamos a asumir que $F$ solo es funcion de $p$ y que
$$q = q(p) $$
De forma que podemos escribir

$$F(x, y, u, p, p(p)) = 0. $$
El gradiente de la superficie $u=u(x,y)$ esta dado por

$$\nabla u = (p, q) = (u_x, y_y). $$
Tomemos un punto fijo $(x_0, y_0)$ y escribamos el plano tangente a la superficie en este punto (en $u=u(x,y)$).

$$p(x- x_0) + q ( y - y_0) - (u - u_0) = 0 \tag{1} $$
Observamos que que la ecuacion la podemos escribir como

$$(p, q, -1) \cdot (x-x_0, y-y_0, u-y_0) =0 $$
Es de la forma

$$\nabla G \cdot (\mathbf{x} - \mathbf{x}_0) $$
donde $\mathbf{x}=(x,y,z)$ y $\mathbf{x}_0=(x_0, y_0, z_0)$
y $G=u - u(x,y)=0$. En palabras $\mathbf{x} - \mathbf{x}_0$ es ortogonal al gradiente $\nabla G$.

Hasta el momento esto es parecido a lo que hicimos con las ecuacion quasilineales. La diferencia es que ahora podemos variar $p$ y obtener muchos planos tangentes. Mediante la variacion de $p$. El gradiente es
$$(p, q(p),-1) $$
La ecuacion (1) produce una **familia** de planos tangente parametrizada por $p$. Esta familia tiene una envolvente (envelope). Esta envolvente es un cono y este cono se llama un **Monge cone**. Tenemos que este cono tiene dos propiedades.

1. Todos los planos comparten $P=(x_0, y_0, z_0)$ que es el apice del cono.
2. El cono no tiene que ser curcular y recto.

En las notas de clase hay un dibujo Figura 2.10 (no la voy a importar).

Las ecuaciones para el Monge conde deben satisfacer:

$$(x-x_0) p + (y- y_0) q = u - u_0 \tag{2} $$
$$ F(x_0, y_0, u_0, p , q(p))=0 \tag{3} $$

Derivamos el plano (2) con respecto a $p$ y encontramos (recuerden que $q=q(p)$.

$$(x-x_0) + (y-y_0) \frac{dq}{dp} = 0. (4) $$

Vamos a asumir que $F_x, F_y$ no se hacen cero simultaneamente (explicamos por que esta hipotesis debe cumplirse en unos momento). Es decir queremos que
$$F_x^2 + F_y^2 \ne 0 . $$
Asumamos por el momento que $F_q \ne 0$.  Tomamos diferenciacion implicita en le ecuacion (3)

$$ F_p + F_q \frac{dq}{dp} = 0 $$
de forma que

$$\frac{dq}{dp} = - \frac{F_p}{F_q} (6) $$

La ecuacion (4) se puede escribir como

$$\frac{x - x_0}{dp/dq} = -(y-y_0) $$
De la (6) aca

$$\frac{x - x_0}{-F_p/F_q} = -(y-y_0) $$
simplificamos (multiplicar a ambos lados por $-F_q$

$$ \frac{x - x_0}{F_p} = \frac{y - y_0}{F_q}$$


Vamos a usar la siguiente propiedad:

Si
$$\frac{a}{b} = \frac{c}{d} $$
entonces
$$\frac{a}{b} = \frac{pa + qc}{pb + qd}$$

Vamos a usar esta igualdad con $a=x-x_0$, $b=F_p$, $c=y-y_0$, $d=F_q$ y encontramos

$$\frac{x - x_0}{F_p} = \frac{y-y_0}{F_q} = \frac{p(x-x_0) + q(y-y_0)}{pF_p + q F_q}$$

[Esta ecuacion genera un cono](http://www.mndcollegerajur.org/uploads/department/cone.pdf)
Piense en el vector $v=(F_p, F_q, pF_p + qF_q)$, todo con parametro $p$ y con apice en el punto $(x_0, y_0, u_0)$. ESta ecuacion es la ecuacion de una recta generadora.

Con lo que hemos hecho escribirmos el **Monge cone** como la superficie en $C=C(x,y)$ generada por las 3 ecuaciones

$$\begin{cases}
F(x_0, y_0, z_0, p, q) = 0 \quad, \quad \text{la ecuacion diferencial} \\
u - u_0 = p(x-x_0) + q (y - y_0 \quad , \quad \text{plano tangente} \\
\frac{x - x_0}{F_p} = \frac{y - y_0}{F_q} \quad , \quad \text{ recta generadora}
\end{cases} \tag{6}  $$
Para encontrar el Monge cone con apice en $(x_0,y_0, z_0=u(x_0, y_0))$
usamos el siguiente ejemplo.

**Ejemplo 2.2.7**: Encuentre el Monge donde de la ecuacion eikonal
que pasa por el punto $(x_0,y_0, u(x_0, y_0)$
En este ejemplo $(x_0, y_0)=(1,1)$

Esta es la ecuacion eikonal
$$ p^2 + q^2 = 1$$
$$F(x_0, y_0, z_0, p, q) = p^2 + q^2 - 1 $$

De aca que
$$ q = \pm \sqrt{1 - p^2}$$

El plano tangente es

$$ u = u_0 + p(x-1) + q(x-1) \tag{7} $$

Ahora bien


$$F_p = 2p  \quad , \quad F_q = 2q $$
Reemplazamos esto en la tercera (recta generadora) de las ecuaciones (6)

$$ \frac{x-1}{2 p } = \frac{y-1}{2q} $$

Necesitamos **eliminar** $p,q$ de las ecuaciones de arriba.
Asumamos que $(x-1)/2p=k$ de forma que

$$ \frac{x-1}{2 p } = \frac{y-1}{2q} = k $$

$$p = \frac{x-1}{2k} \quad , \quad q= \frac{y-1}{2k} $$
Sustituimos estas ecuaciones en la ecuacion del plano tangente (7)

$$ u = u_0 + \frac{(x-1)^2}{2k} + \frac{(y-1)^2}{2k} = u_0 + \frac{(x-1)^2 + (y-1)^2}{2k} \tag{8} $$

Como $p^2 + q^2 = 1$,

$$\frac{(x-1)^2}{4k^2} + \frac{(y-1)^2}{4 k^2} = 1 $$
$$(x-1)^2 + (y-1)^2 = 4 k^2$$

Ahora bien  de la (8)
$$(x-1)^2 + (y-1)^2 = 2k(u - u_0) $$
Comparando las ultimas dos ecuaciones tenemos que

$$4 k^2 = 2k(u - u_0) $$
de donde

$$u = 2k + u_0 $$

Entonces

$$ (x- 1)^2 + (y-1)^2 = 2k ( u- u_0) =(u-u_0)^2$$

Ahroa, no sabemos quien es $u_0$ ? Enseguida mostraremos que la solucion de la ecuacion eikonal es $x^2 + y^2 = u^2$, de forma que si $(x,y)=(1,1)$ entonces
$u_0 = \sqrt{1^2 + 1^2} = \sqrt{2}$ con lo que
el Monge cone es

$$(x-1)^2 + (y-1)^2 = (u - \sqrt{2})^2 $$







```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define cylindrical coordinates
r = np.linspace(0, 2, 50)
theta = np.linspace(0, 2*np.pi, 50)
R, Theta = np.meshgrid(r, theta)
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

# Define the two cones
Z1 = np.sqrt(X**2 + Y**2)  # First cone: u^2 = x^2 + y^2
Z2 = np.sqrt((X-1)**2 + (Y-1)**2) + np.sqrt(2)  # Second cone: (x-1)^2 + (y-1)^2 = (u - sqrt(2))^2

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the cones
ax.plot_surface(X, Y, Z1, alpha=0.5, label='Cone 1')
ax.plot_surface(X, Y, Z2, alpha=0.5, label='Cone 2')

# Label the apices
ax.scatter(0, 0, 0, color='red', marker='o', label='Apex of Cone 1')
ax.scatter(1, 1, np.sqrt(2), color='blue', marker='o', label='Apex of Cone 2')

# Label the line of intersection (This is an approximation)
# The actual intersection line is more complex and requires numerical methods
# to compute accurately.
# This might be slightly inaccurate, but try the following:
# ax.plot([0, 1], [0, 1], [0, np.sqrt(2)], color='black', linestyle='--', label='Intersection Line')


# Set plot labels and legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u')
ax.legend()

# Display the plot
plt.show()
```

Retornamos al problema de la solucion de la ecuacion no lineal de primer orden (parcial)

$$a(x,y,u) p + b(x,y,u) q = c(x,y,u)  $$

En este caso la funcion $F$ es

$$F(x,y,u, p, q)= a(x,y,u) p  + b(x,y,u) q  - c(x,y,u) = 0 $$

Como en el metodo tradicional de las caracteristicas pensamos en el vector
$$v = [a(x,y,u), b(x,y,u), c(x,y,u) $$
y recordando la ecuacion de la linea guia (guiding line)

$$\frac{x-x_0}{F_p} = \frac{y-y_0}{F_q} = \frac{u-u_0}{p F_p + q F_q} \tag{9} $$


En este caso como $F_p=a(x,y,u)$, $F_q=b(x,y,u)$,
ademas

$$p F_p + q F_q = a(x,y,u) p + b(x,y,u) q = c(x,y,u) $$
entonces tenemos

$$\frac{x-x_0}{a(x,y, u)} = \frac{y-y_0}{b(x,y,u)}  = \frac{u-u_0}{c(x, y, u)} $$
Esta es la linea tangente que guia al cono.

Usando diferenciales en (9) (cuando $x \to x_0$, $y \to y_0$) podemos escribir

$$\frac{dx}{F_p} = \frac{dy}{F_q} = \frac{du}{p F_p + q F_q} $$

O de otra forma si parametrizamos las curvas con $\sigma$,
$(x,y,u)= (x(\sigma, \tau)), y(\sigma, \tau)), u(\sigma, \tau))$

$$ \frac{d x(\sigma)}{d \sigma} = F_p \quad , \quad  \frac{d y(\sigma)}{d\sigma} =  F_q \quad , \quad \frac{d u(\sigma)}{d \sigma} = p F_p + q F_q  \tag{9} $$

Estas son las ecuaciones caracteristicas. Pero hay un problema. Tenemos dos nuevas variables $p,q$. Como las eliminamos?

Comparando con las ecuaciones quasilineales vemos mucha similitud.
El problema es que las no lineales, no son lineales en $p=u_x$ o en $q=u_y$.

Que pasa si
$$F_p^2 + F_q^2 = 0?$$
Piense en la ecuacion $p^2 + q^2=1$, $F(x,y,u,p,q)=p^2 + q^2 -1 = 0$.
$F_p=2p, F_q=2q$, y el lado derecho de la ecuacion (9) se convierte en

$$ p F_p + q F_q = 2 (F_p^2 + F_q^2) = 0 $$


**Ejemplo 2.2.8**: Asuma la ecuacion PDE no lineal $u_x^2 + u_y^2 = u$, resuelvala

Solucion, en terminos de $p,q$
$$ F(x,y,u,p,q) = p^2 + q^2 - u = 0.$$
De aca que $F_p=2p$, $F_q=2q$,  las caracteristicas  (9) se reducen a

$$ \frac{d x(\sigma)}{d \sigma } = 2 p   \quad, \quad
 \frac{d y(\sigma)}{d \sigma } = 2 q   \quad , \quad
 \frac{d u(\sigma)}{d \sigma } = 2 u  
$$

Aparentemente ya tenemos todos los elementos para resolver el problema. Pero
**Quien es $p$? Quien es $q$?**
En la proxima clase, construirmos **dos** ecuaciones mas para poder resolver por $p,q$. O sea que necesitariamos **5 ecuaciones** en las variables
$x,y,p,q,u$.
