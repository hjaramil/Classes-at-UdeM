### Derivadas de distribuciones
La **derivada** de una distribucion $T$ se define por la forma funcional

$$\langle T', \varphi \rangle = -\langle T , \varphi' \rangle  $$
Para cualquier funcion de prueba $\varphi$. Esto verificar usando integracion por partes. Lo dejamos de ejercicio al estudiante.

Observamos la conveniencia de la definicion de distribucion.

1. Los bordos de $\varphi$ son 0 y sus derivadas. Entonces la integracion se realiza sobre el dominio limitado (acotado y cerrado). (compacto).

2. Usted puede hacer integracion por partes arrojando la derivada de $T$ en la derivada de $\varphi$ que es infinitamente diferenciable.

De esta forma podemos extender la derivada a cualquier orden que queramos.

$$\langle T^{(n)}, \varphi \rangle = (-1)^{n} \langle T, \varphi^{(n)} \rangle  $$
Tambien se prueba esto usando integracion por partes.

Un ejemplo clasi es la derivada del Delta de Dirac, o sea $\delta'(x)$.


$$\langle \delta'(x), \varphi(x) \rangle = -\langle \delta(x), \varphi'(x) \rangle = - \varphi'(0) $$

Igualmente

$$\delta^{(n)} $$
es tal que

$$\langle \delta^{(n)}(x), \varphi(x) \rangle = (-1)^n \langle \delta(x), \varphi^{(n)}(x) \rangle = (-1)^n \varphi^{(n)}(0) $$


Vamos encontrar tres formas de hallar la derivada de una funcion descontinua importante. Esta es la Heaviside function

$$H(x) =
\begin{cases}
0 \quad , \quad x < 0 \\
1/2 \quad , \quad x = 0 \\
1 \quad , \quad x> 0
\end{cases}
 $$

```python
# plot the heaviside function
import numpy as np
import matplotlib.pyplot as plt

def heaviside(x):
    return np.where(x < 0, 0, np.where(x == 0, 0.5, 1))

# graph the function
x = np.linspace(-2, 2, 100)
y = heaviside(x)
plt.plot(x, y)
plt.show()
```

Vamos a hallar la derivda de $H$ de cuatro formas distintas y probar que es $\delta(x)$.

1. Usando la definicion de derivada (sentido clasico) con el concepto de limite en $H$
Por fuera de $0$ $H(x)=0$ y la derivada es 0. El problema es como hallamos la derivda en $x=0$.

Limite por la derecha
$$\lim_{h \to 0^+} \frac{H(0 + h) - H(0)}{h} =  \frac{1/2}{0} = +\infty$$

Limite por la izquierda
$$\lim_{h \to 0^-} \frac{H(0 + h) - H(0)}{h} =  \frac{0 - H(0)}{0} = \frac{-1/2}{0}=  -\infty$$

La derivada en el sentido **clasico** no existe por dos razones.

* No hay valor finito en $x=0$.
* El limite por la izquierda es diferente del limite por la derecha.


2. Veamos como, usando distribuciones, podemos encontrar $H'(x)$.

$$\langle H', \varphi \rangle = -\langle H, \varphi'\rangle
=- \int_0^{\infty} \varphi'(x) dx = -(\varphi(\infty) - \varphi(0))= \varphi(0)  $$
Pero recuerde la definicion de delta de Dirac

$$\delta(x) = \varphi(0) $$
O sea que que

$$H'(x) = \delta(x).  $$

3. Definamos

$$f(x) = \int_{-\infty}^x \delta(t) dt  \tag{1} $$

Entonces

$$
f(x) =
\begin{cases}
1 \quad , \quad \text{ si  } x > 0  \\
0 \quad , \quad \text{ si  } x < 0  \\
\end{cases}
$$
Esto es la Heaviside function. Ahora quien es $f'(x)$. Usando el teorema fundamental del calculo en (1).

$$f'(x) = \delta(x)$$.

es decir
$$\frac{d H(x)}{d x} = \delta(x) . $$

4. Volvemos a la definicion con limite pero usando distribuciones.


$$\left \langle \frac{1}{h} [ H(x + h) - H(x) , \varphi \right\rangle=
\frac1h  \left ( \int_{-h}^\infty  \varphi(x) dx - \int_0^\infty \varphi(x) dx  \right ) = \frac1h \int_{-h}^0 \varphi(x) dx = \varphi(0). $$

Por definicion de derivada encontramos que

$$\langle H'(x), \varphi(x) \rangle = \varphi(0). $$

Una vez mas

$$\frac{d H(x)}{dx} = \delta(x). $$


Con esta informacion mostramos como sacar la derivada de una funcion con un salto.

Piense en que en el punto $x_0$ la funcion $f(x)$ se rompe sumandole un **salto** que se $\Delta f$.

Es decir

Asumimos que $f(x)$ es continua y diferencialble en todas partes.
$$g(x) =
\begin{cases}
f(x) \quad , \quad \text{si } x < x_0 \\
f(x)  + \Delta f \quad , \quad \text{si } x > x_0 \\
\end{cases} $$

Queremos hallar la derivada de $g$. Para esto escribimos $g$ de la siguiente forma

$$g(x) = f(x) + \Delta f H(x -x_0). $$

Usando distribuciones sacamos la derivada de $g$.

$$g'(x) = f'(x)  + \Delta f \delta(x-x_0). $$

Generalizamos este ejemplo a un numero finito de saltos. Es decir, asumamos que $f$ es continua y diferenciable entonces para un numero $n$ de saltos tenemos

$$g(x) = f(x)  + \sum_{i=0}^n \Delta f_i H(x - x_i) . $$

Si sacamos la derivada de esto encontramos

$$g'(x) = f'(x) + \sum_{i=0}^n \Delta f_i \delta(x - x_i). $$

Otro ejemplo, la funcion signo.

$$
\text{sgn}(x)  =
\begin{cases}
-1 \quad , \quad \text{si } x < 0 \\
0 \quad , \quad \text{si } x = 0 \\
1 \quad , \quad \text{si } x > 0 \\
\end{cases}
$$

Tarea mostrar que

$$\text{sgn}(x) = 2 H(x) - 1. $$

De aca que

$$\frac{d \text{sng}(x)}{dx} = 2 \delta(x).  $$

### Otras opearaciones en distribuciones.
Si $T$ es una distibucion y $g$ es una funcion suave. Definimos la multiplicacion $g(x) T$ con la forma funcional.

$$\langle g T, \varphi \rangle = \langle T, g(x) \varphi(x)  \rangle  $$

#### Convolucion.
La convolucion de una distribucion $T$ con una funcion de prueba $\varphi$ se define de la forma clasica

$$(T * \varphi)(x) = \int_{-\infty}^\infty  T(x-y) \varphi(y) dy  $$












## Aplicaciones de todo lo que hemos visto en distribuciones.

* **Ecuaciones diferenciales**: Las distribuciones permiten la solucion de ecuaciones diferenciales con fuentes. Una fuente importante es un Delta de Dirac. Que es una fuente puntual.  Por ejemplo, veremos que la funcion de Green es una respuesta a una fuente puntual (delta de Dirac). A esto se le llama **respuesta impulso** (impulse response).

* **Tranformadas de Fourier**: La transformada de Fourier se extiende a distribuciones mostrando como se pueden resolver las PDF con este metodo.

* **Fisica**: Un fenomeno fisico, como la onda, se puede ver como una respuesta a cargas (fuentes) puntuales superpuestas. Como veremos hoy.



# Funciones de Green (Green's functions).
Las funciones de Green son herramientas matematicas usadas para resolver ecuaciones diferenciales con fuente. La funcion de Green representa la solucion a una ODE o PDE (ordinaria o parcial) con fuente puntual (Dirac Delta). Es decir es la respueta impulso del sistema.


## Motivacion:
Vamos a comparar la solucion de sistemas lineales algebraicos con los sitemas lineales diferenciales.

Es decir, piense en el sistema

$$Ax = b \tag{2} $$
donde $A \in \mathbb{R}^{n \times n}$ y $x,y$ son vectores en $\mathbb{R}^n$.
Asumimos ademas que $A^{-1}$ existe.

Si $A^{-1}$ existe, podemos resolver la Ecuacion (2)  para cualquier "fuente" $b$.  Por ejemplo si conocemos $A^{-1}$ y la "fuente" es $y$ entonces la
ecuacion $A x = y$ se resuelve como

$$x = A^{-1} y. $$

De otro lado tenemos operadores de la forma


$$L u = f \tag{3}  $$
donde $L$ es un operador lineal (diferencial, una ODE o una PDE)
quremos resolver esta ecuacion. Encontrar $u$.

Vamos a retornar a la ecuacion (2) con distintas "fuentes"


$$A x_1 = \mathrm{e}_1 $$
$$A x_2 = \mathrm{e}_2 $$
$$ \vdots $$
$$A x_n = \mathrm{e}_n \tag{4} $$
donde $\mathrm{e}_i$ es un vector canonico con 0 en todas partes menos en la componente $i$. O sea $(\mathrm{e}_i)_j = \delta_{ij}$ .

Construyamos ahora la ecuacion

$$A [x_1 | x_2 | \cdots | x_n ] = [ \mathrm{e}_1 | \mathrm{e}_2 | \cdots | \mathrm{e}_n]  = I $$

De forma que tenemos el sistema

$$AB = I $$
donde
$$ B= [x_1 | x_2 | \cdots| x_n ]$$
Claramente $B=A^{-1}$.

En notacion indizal las ecuaciones (4) las podemos escribir como

$$A( x_i)_j = \delta_{ij}  \tag{5} $$

Vamos a extender esta idea al operador (4). Queremos hallar algo como

$$L G(x, \xi) = \delta(x - \xi) \tag{6}$$

Comparamos las ecuaciones (5) con (6).

* La (5) es en el discreto finito, la (6) en el continuo infinito.
* $$ A ↔  L $$
* $$ G ↔  A^{-1} $$
* $$ i ↔  x $$
* $$ j ↔  \xi $$
* $$ \delta_{ij} ↔  \delta(x- \xi) $$
