## Tensores.

**Motivaci√≥n**:
Piense en un conjuno de im√°genes. Las im√°genes tienen pixeles ($p,q$) horizontal/vertical. Pero cada pixel puede tener 3 o 4 canales (RedGreenBlue).
Para describir la amplitud de un punto en una imagen especifica se necesitan 4 indices
* $i$: N√∫mero de im√°gen
* $j$: pixel en $x$
* $k$: pixel en $y$
* $l$: canal.
La amplitud (intensidad) est√° dada por
$a_{ijkl}$.


Este objeto no se puede describir ni con vectores, ni con matrices.
Este es un objeto **multi-index** (tensor de "dimension" 4. en CS).

Una forma elegante, matem√°tica de ver tensores es como **transformaciones multilineales**. El objeto de esta sesi√≥n es solo ilustraci√≥n y un poco de formalismo, pero no se aplica a nada de lo que vemos de DL.

## Transformaciones multi-lineales.
El m√©todo para explicar esto es de abajo hacia arriba, de simple a complejo.

* Asumamos $x \in \mathbb{R}$, $y \in \mathbb{R}$, $a \in \mathbb{R}$.
Tefinimos una transformaci√≥n lineal como

$$T x = y = a x $$

Esta la escribo formal, las otras no

\begin{eqnarray}
T : \mathbb{R} &\to& \mathbb{R} \\
x &\mapsto& Tx = ax
\end{eqnarray}
Por que es lineal

$$T(\alpha x + \beta y) = \alpha T(x) + \beta T(y) = a \alpha x  + \beta a x $$

Por que $T(a x) = a x + b$
$$T(\alpha x ) = \alpha a x + b \ne \alpha T(x)  $$
Afine transformation.


Ac√° $y$ es un objeto (tensor) de **rango 0**.

* Asumamos ahora que $x \in \mathbb{R}^n$ , $y \in \mathbb{R}^m$
Queremos

$$T x = y = ax $$
Que objeto es $a$?  Del algebra que $a \in \mathbb{R}^{m \times n}$
$y$ es un **tensor de rango 1**.

Podemos usar notaci√≥n indizal y referir a $y_i$ como la componente $i$ del vector $y$.  Observe que $T$ es lineal en $x$, es decir para $c_1, c_2 \in \mathbb{R}$

$$T( c_1 x_1 + c_2 x_2) = c_1 T(x_1) + c_2 T(x_2) $$
Si asumimos la base can√≥nica en $\mathbb{R}$

\begin{eqnarray}
e_1 = \begin{pmatrix} 1 \\ 0 \\ \vdots \\ \vdots \\ 0  \end{pmatrix}  \quad, \quad
e_2 = \begin{pmatrix} 0 \\ 1 \\ \vdots \\ \vdots \\ 0  \end{pmatrix}  \quad, \quad
e_k = \begin{pmatrix} 0  \\ \vdots \\ 1 \\ \vdots \\ 0  \end{pmatrix}  \quad, \quad
e_n = \begin{pmatrix} 0  \\ \vdots \\ 0 \\ \vdots \\ 1  \end{pmatrix}  \quad, \quad
\end{eqnarray}

Podemos escribir

$$x = \sum_{i=1}^{n} x_i e_i. $$
entonces

$$T x = T \left ( \sum_{i=1}^n x_i e_i \right ) = \sum_{i=1}^n x_i T(e_i)  $$
donde la $j$ columna de la matriz $a$ esta dada por $T(e_j)$.

$$a = \begin{pmatrix}
| & | & \vdots & \vdots & | \\
T(e_1) & T(e_2) & \vdots & \vdots & T(e_n)  \\
| & | & \vdots & \vdots & | \\
\end{pmatrix} $$

* $x \in \mathbb{R}^{n \times m}, y \in \mathbb{R}^{p \times q}$
$$ Tx = a x$$
quien es $a$?
Debemos hallar las dimensiones de $a$. Necesitamos hallar las **bases**
en el espacio de entrada $\mathbb{R}^{ n \times m}$. Si por ejemplo $m=n=2$,
las bases (matrices) de esta espacio son

\begin{eqnarray}
e_{11} = \begin{pmatrix} 1 & 0 \\ 0 & 0  \end{pmatrix}  \quad , \quad  
e_{12} = \begin{pmatrix} 0 & 1 \\ 0 & 0  \end{pmatrix}  \quad , \quad  
e_{21} = \begin{pmatrix} 1 & 0 \\ 0 & 0  \end{pmatrix}  \quad , \quad  
e_{22} = \begin{pmatrix} 0 & 0 \\ 0 & 1  \end{pmatrix}  \quad , \quad  
\end{eqnarray}
Vamos a definir el **producto tensorial** (o **outer product**, **producto externo**. $c = a \otimes b$ si

$$c_{ij} = (a \otimes b)_{ij} = a_i b_j $$
Entonces, por ejemplo

$$e_{11} = e_1 \otimes e_1  =
\begin{pmatrix}
1 \times 1  & 1 \times 0 \\ 0 \times 1 & 0 \times 0
\end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}$$

Es claro que si, por ejemplo en $\mathbb{R}^{2 \times 2}$
$$ \begin{pmatrix} a & b \\ c & d \end{pmatrix} = a e_{11} + b e_{12} + c e_{21} + d e_{22} $$






image.png

Podemos escribir

$$x = \sum_{i=1}^n \sum_{j=1}^{m}  x_{ij} e_{ij} = \sum._{i=1}^n \sum_{j=1}^m x_{ij} e_i \otimes e_j  $$
Entonces aplicando el operador lineal $T$

$$T(x) = \sum_{i=1}^n \sum_{j=1}^m x_{ij} T(e_{ij}) $$
El objeto $T(e_{ij})$ para $i,j$ fijos es un obejto (matriz) en
$\mathbb{R}^{p \times q}$. Podemos usar indices para escribir

$$T_{st}(x) = \sum_{i=1}^n \sum_{j=1}^m x_{ij} T_{st} (e_{ij}) \quad , \quad s=1,2 \cdots, p \quad , \quad t=1,2, \cdots, q $$
El objeto $a$ tiene la forma

$$a_{st}^{ij} = T_{st}(e_{ij}) $$
Esto objeto $a$ tiene 4 indices. Los indices de suma $i,j$ se llaman **dummies**
(mudos) y los indices $s,t$ se llaman **free** (libres). La suma se hace **siempre** sobre los **dummies**.

Es costumbre evitar los s√≠mbolos de suma. A esto se le llama la **notaci√≥n de Einstein**.
[Teor√≠a de la relatividad general](http://www.alberteinstein.info/gallery/gtext3.html). Con esta notaci√≥n se escribe

$$T_{st}(x) = x_{ij} T_{st}(e_{ij}) = a_{st}^{ij} x_{ij}  $$

Un producto interno se escribe $y^i x_i$

* Vamos a generalizar mucho mas.
Que tal si
$$x \in \mathbb{R}^{n_1 \times n_2 \times \cdots \times n_j}  $$
$$y \in \mathbb{R}^{m_1 \times m_2 \times \cdots \times n_k}  $$
La pregunta es
$$Tx = a x = y  $$
quien es $a$?

De igual forma que antes

$$e_{p_1 p_2 \cdots p_j} = e_{p_1} \otimes e_{p_2} \cdots \otimes{e_{p_j}}$$.

Usando la notaci√≥n de Einstein

$$x = x_{p_1 p_2 \cdots p_j} e_{p_1 p_2 \cdots p_j} $$
y
$$Tx = x_{p_1 p_2 \cdots p_j} T(e_{p_1 p_2 \cdots p_j}) $$
donde
$T(e_{p_1 p_2 \cdots p_j}) \in \mathbb{R}^{m_1 \times m_2 \times \cdots m_k }  $
y

$$\left [ T(e_{p_1 p_2 \cdots p_j})   \right ]_{q_1 q_2 \cdots q_k} = a_{q_1 q_2 \cdots q_k}^{p_1 p_2 \cdots p_j}  $$

De forma que podemos escribir

$$T_{q_1 q_2 \cdots q_k} (x) = a_{q_1 q_2 \cdots q_k}^{p_1 p_2 \cdots p_j} x_{p_1 p_2 \cdots p_j} $$


Entones
$T_{q_1 q_2 \cdots q_k} (x)$ es un tensor de rango $k$.

Si $q_1$ va de 1 a $n_1$
Si $q_2$ va de 1 a $n_2$
Si $q_k$ va de 1 a $n_k$
La dimension de este espacio es
$$\prod_{i=1}^k q_i $$

Por ejemplo si una pelicula con 10 marcos cada uno con $400 \times 600$ pixeles y 3 canales tiene

$$10 \times 400 \times 600 \times 3 = 7200000.$$
