### Matriz Adjunta
Sea $A$ una matriz, la matriz conjugada de $A$ se nota $B=\overline{A}$ es aqueya tal que $b_{ij} = \overline{a_{ij}}$. Reuerden la definicion de **conjugado**. Si $x = a + \mathrm{i} b$ donde $a, b \in \mathbb{R}$ y $\mathrm{i} = \sqrt{-1}$, entonces el conjugado de $x$ es $\overline{x}=a - \mathrm{i} b$. El conjugado de un numero complejo es la reflexion con respecto al eje real. La matriz **adjunta** de $A$ que se escribe $B=A^*$ 
es aqueya donde $b_{ij} = \overline{a_{ij}}$ es decir 

$$ A^* = (\overline{A})^T = \overline{A^T}$$\
En palabras la la matriz adjunta de $A$ es la transpuesta, conjugada de $A$. 


\begin{eqnarray}
A = \begin{pmatrix} 3 + 3 \mathrm{i} & 1 - \mathrm{i} \\
5 & 3 - 4 \mathrm{i} 
\end{pmatrix} \\
A^* = \begin{pmatrix} 3 - 3 \mathrm{i} & 5 \\
1 + \mathrm{i} & 3 + 4 \mathrm{i} 
\end{pmatrix}
\end{eqnarray}

### Producto interno.
Sean $u, v \in \mathbb{C}^n$, definimos el producto interno (inner product) de $u$ con $v$ como $u^* v$. La notacion:

$$\langle u, v \rangle = u^* v $$

### Ortogonalidad
Dos vectores $a,b$ son **ortogonales** si su producto interno es 0.
Es decir si $\langle a, b \rangle =0$. Decimos que una matriz $A$ es **ortogonal** si sus columnas son ortogonales entre si. Decimos que es **ortonormal** si ademas las columnas tienen norma 1. La norma la definimos como.

$$ \| u \| = \sqrt{ \langle u, u \rangle} $$
La palabra ortonormalidad en matrices no se usa. Se dice que son ortogonales. Si la matriz es $U=[u_1 | u_2 | \cdots u_n ]$, entonces
$\langle u_i , u_j \rangle $ es 0 si $i \ne 0$, y $1$ si $i=j$.
Es decir es un delta de Kronecker.

$$ 
\langle u_i, u_j \rangle = \delta_{ij} = \left \{
    \begin{array}{cc}
    1 & \text{si  } i = j \\
    0 & \text{si } i \ne j 
    \end{array}
    \right .
$$

Por que la norma se define como la raiz cuadrada del producto interno.
Piense en $\mathbb{C}^2$.  $u = (3 + 2 \mathrm{i}, 1 - \mathrm{i})^T$, 

La norma de $u$ es

$$\| u \| = \sqrt{\langle u, u \rangle} = \sqrt{u^* u  } =
\sqrt{ (3 - 2 \mathrm{i} , 1 + \mathrm{i} ) \begin{pmatrix} 3 + 2 \mathrm{i} \\ 1 - \mathrm{i}
\end{pmatrix}} = \sqrt{ 3^2 + 2^2  + 1 +1 } = \sqrt{15}$$

#### Propiedades la matriz adjunta.
Dadas dos matrices $A,B$. Entonces (teorema)

* La matriz adjunta del producto es el producto (reverso) de las matrices adjuntas. En simbolos

$$(AB)^* = B^* A^*  $$
* Ademas:

\begin{eqnarray}
(A \pm B )^* &=& A^* \pm B^* \\
(\alpha A)^* &=& \overline{\alpha} A^*  \\
(A^*)^* &=& A \\
\overline{AB} &=& \overline{A} \overline{B} 
\end{eqnarray}

#### Propiedades del producto interno: Teorema.
Si pensamos en producto interno en los reales $\mathbb{R}$ (producto punto) el producto interno es **conmutativo**. Tarea, buscar un contra-ejemplo en los complejos donde $\langle u, v \rangle \ne \langle v, u \rangle $. 

* Exaccion del escalar del producto interno. 
\begin{eqnarray}
\langle \alpha u , v \rangle &=& \overline{\alpha} \langle u, v \rangle \\
\langle u, \alpha v \rangle &=& \alpha \langle u, v \rangle 
\end{eqnarray}

* propiedad **distributiva** del producto interno

\begin{eqnarray}
\langle u, v \pm w \rangle &=& \langle u , v \rangle \pm \langle u, w \rangle  \\
\langle u \pm v , w \rangle &=& \langle u , w \rangle \pm \langle v, w \rangle
\end{eqnarray}

Del teorema de Pitagoras:

$$ \| x \|^2 + \| y \|^2 = \| x + y \|^2  \quad (1) $$

Pero

$$\| x + y \|^2 = \langle x + y , x + y \rangle = \langle x, x \rangle + \langle x, y \rangle + \langle y , x \rangle + \langle y, y \rangle = \| x \|^2 + 2 \langle x, y \rangle + \| y  \|^2 
\quad (2)   $$
Si comparamos (1) con (2) ( o mejor, restando (1) de la (2)) encontramos que $2 \langle x , y \rangle = 0$, es decir $\langle x, y \rangle = 0$.
O sea que $x$ es ortogonal a $y$. 

El teorema de pitagoras se puede probar de forma "contraria". Es decir

Definicion (rango y el espacio nulo (kernel): Dada una matriz $A$, $m \times n$ el **rango** de la matriz es el espacio vectorial generado or las columnas de la matriz. Por esto tambien se le conoce como el espacio columna. Es decir si la matriz $A$ tiene columnas 

$A = [A_1 | A_2 | \cdots A_n]$ el espacio columna (o rango) es el generado por todas las columnas.

$$\mathcal{R}(A) = \left \{  \sum_{i=1}^n \alpha_i A_i  \mid  \alpha_i \in \mathbb{C} \right \} $$
El espacio **nulo** o **kernel** es el espacio de todos los vectores $x$ tales que $Ax = 0$.

En matematicas:

$$\mathcal{N} = \left \{  x | A x = 0   \right \} $$



## Autovalores y autovectores.
Dada una matriz $A$ cuadrada, si existe $\lambda \in \mathbb{C} $  y $x \ne 0$, tal que
$$ Ax = \lambda x  \quad (5) $$
decimos que $\lambda$ es un autovalor de $A$ y que $x$ es el autovector correspondiente al autovalor $\lambda$. 

Cual es el concepto? esto que es? para que sirve?

[Wikipedia](https://en.wikipedia.org/wiki/Eigenvalues_and_eigenvectors) muestra ejemplos con animaciones ( y sin animaciones) ilustrativos del concepto de autovalor y autovector. 

Como se hallan los autovalores, autovectores?
De la ecuacion (5) pasando a restar y como $x = Ix$.

$$ Ax - \lambda Ix  = 0  $$
factorizamos $x$ factor comun a derecha.

$$(A - \lambda I) x = 0  $$
Del algebra lineal hay un teorema que dice que si $Bx = 0$ y $x$ no es 0 entonces $\det B = 0$.
Entonces

$$ \det (A - \lambda I) = 0 $$
A esta ecuacion se le llama la ecuacion **caracteristica** de la matriz $A$. Es un polinomio en $\lambda$.


$$ p(\lambda) = \det (A - \lambda I) = 0 $$

Ejemplo. Halle el polinomio caracteristoc de la matriz

$$A =  \begin{pmatrix} 3 & 4 \\ 5 & -1 \end{pmatrix} $$

Solucion. Hallar el determinante de $A - \lambda I$.

$$ \det (A -\lambda I) = \det  \begin{pmatrix} 3 -\lambda & 4 \\ 5 & -1 - \lambda \end{pmatrix} $$

Es decir

$$ p(\lambda) = (3-\lambda)(-1-\lambda) - 20 = -3 - 3\lambda + \lambda + \lambda^2 = \lambda^2 -2 \lambda - 3 $$
