# Metodo de steepest descent (maximo descent)

$$ J = \frac12 \| h(x, \theta) - b \|^2$$
donde $h(x, \theta) = A \theta$, con $A  = X^T X$, $b=X^T y$.

El gradiente $J$ se puede calcular con las notas de **matrix calculus** (clase dos antes de esta). Encontramos alli:

$$ \nabla J = X^T (X \theta - y) = A \theta - b \quad (1) $$, fijense que si $\nabla J=0$
tenmos las famosas **ecuaciones normales**

$$ X^T X \theta = X^T y  \quad \text{o} \quad A \theta = b $$

Como se **itera** para llegar al minimo por costo. Invertir $X^T X$ no es facil.

$$ \theta_k = \theta_{k-1} - \alpha \nabla J(\theta_{k-1}) $$

La pregunta en este momento es quien es $\alpha$. A este se le llama "paso".
Si el pasos es pequeno, casi que garantizamos convergencia pero a un "paso" muy lento. Si es muy grande el sistema podria diverger.



Selection_003.png

El tamano del paso $\alpha$ es un problema delicado. Veamos por ejemplo el caso de SD: Steepest Descent. A la busqueda de $\alpha$ se le llama **line search algorithm**. Veamos el algoritmo de busqueda de paso en SD.

Debemos minimizar la funcion objetivo $J$. Asumamos que estamos al comienzo de la iteracion $\theta_0$, y queremos encontrar $\theta_1 = \theta_0 - \alpha \nabla J(\theta_0)$. Podemos definir una funcion auxiliar:

$$ \varphi(\alpha) =  J[\theta_0 - \alpha \nabla J(\theta_0)] $$
Podemos pensar en minimizar $\varphi(\alpha)$ como funcion de $\alpha$, pero note que esta es una funcion de una variable. Si tomamos la derivadad, usamos la regla de la cadena y la igualamos a 0 encontramos:

$$ \varphi'(\alpha) = \nabla J[ \theta_0 - \alpha \nabla J(\theta_0)] \cdot \nabla J(\theta_0)  = 0  \quad (2) $$

Debemos resolver esta ecuacion para $\alpha$. NO se ve facil. Toca hacer trucos, y estos trucos son el alma del metodo de SD.

$$ \theta_1 = \theta_0 - \alpha \nabla J(\theta_0) \quad (4)  $$

Definamos los **residuales** 

$$ r_0 = A \theta_0 - b = \nabla J(\theta_0) $$ por la ecuacion (1)
$$ r_1 = A \theta_1 - b = \nabla J(\theta_1) \quad (3) $$

De la ecuacion (2) vemos que $r_0 \cdot r_1 = 0$, es decir $r_0 \perp r_1$.

Hasta ahora no hay truco, simplemente es redefinir los terminos para mejor manipulacion. El truco es multiplicar la ecuacion (4) a izquierda por $A$ y restamos $b$


$$ A \theta_1 - b = A \theta_0 - \alpha A \nabla J(\theta_0) - b$$
y por (3) obtengo

$$ r_1  =  r_0 - \alpha A r_0 $$
multiplicamos esta ecuacion por $r_0^T$ a izquierda

$$ r_0^T r_1  = r_0^T r_0 - \alpha r_0^T A r_0 $$

Es decir

$$ 0 = \| r_0 \|^2 - \alpha r_0^T A r_0$$

De forma que

$$ \alpha = \frac{\| r_0 \|^2}{r_0^T A r_0 } $$
El $\alpha$ optimiza el descenso.

Queda la pregunta si podemos bajar apuntando mejor. En un plano si podemos hacer eso. Bajar directamente al minimo, pero la direccion no puede ser ortogonal al contorno. No es la del gradiente. La respuesta es **si**. El metodo se conoce como **conjugate gradient** y fue desarrollado en 1952 simultameante por Hestenes, Stiefel y Lanczos.

La idea es que existe una direccion llama **conjugada** y esa direccion apunta (en un plano) direrctamente al minimo en ese plano. 

Se define un concepto nuevo de ortonalidad llamado ortogonalidad $A$. Two vectors $u$, $v$ are $A$ orthogonal si

$$ u^T A v = 0 $$
Las direcciones $u$ y $v$ se llaman conjugadas. Y si en
un plano usted esta apuntando a $u$, la proxima direccion que lo lleva al centro es la $v$.

```python
import numpy as np
import matplotlib.pyplot as plt

# generar datos

np.random.seed() # los datos aleatorios son buenas para pruebas
mu, sigma=0,1
N=5 # 5 muestras solo para entender, luego 1000
Z = np.random.normal(mu, sigma, N)

phi=0.3
X = np.zeros(N)
X[0]=Z[0]

# Modelo AR: Autor-regresivo
for i in range(1,N):
    X[i] = Z[i] + phi*X[i-1]

plt.xlabel("sample number")
plt.ylabel("sample value")
plt.title("proceso AutoRegresivo AR(1), $\phi=0.3$")
plt.plot(X)
plt.show()


```

Removemos las dependendencias lineales con el lag $k=$ ($X_{t-3}$). Y ver que efecto tiene.
El plan de trabajo es:

* Hallar la regresion lineal multivariadad 
* Remover la dependencia con $X_0, X_1, X_2$.



```python
# regresion lineal multivariada
ones = np.ones( shape=(n,1), dtype='float')
# stack ones
Xmat = np.column_stack( (ones, Y[1]))
for i in range(2, k-1):
    Xmat = np.column_stack((Xmat, Y[i]))

XT= Xmat.transpose()
XTX = XT.dot(Xmat)
XTXInv = np.linalg.inv(XTX)
b =np.matmul(Xmat.transpose(), Y[0])
theta = XTXInv.dot(b)

# ahora vamos substraer de $X$ la contribucion de las otras
# regresiones lineales para obtener los residuales
# substraer sum( theta[j]*Y[j] for j in range(1, k-1))

Y_t = X - theta[0] - sum( theta[j]*Y[j] for j in range(1, k-1))
Y[k-1] = Y_t

# Graficamos los resultados
data = np.transpose(Y)
df = pd.DataFrame(data)


X1 ="$X_t$"
X2 = "$X_{t-1}$"
X3 = "$X_{t-2}$"
X4 = "$X_{t-3}$"

df.columns = [X1, X2, X3, X4]
sns.pairplot(df)
```

## feature normalization : normalizacion de atributos

Que ventaja tiene la normalizacion de los atributos

1. Incrementar la velocidad de convergencia.
2. Incrementar la precision, por que si las columnas de la matriz $X$ estan bien balanceadas, los valores singulares estan mas pegados (unos con otros en tamano) y el numero de condicion $\kappa$ es mas pequeno y por lo tanto mayor precision.
3. Crea un modelo mas significativo. 

si normalizamos las columnas, entonces cambia la solucion?, si cambia pero el problema que queremos resolver no.

Por ejemplo: el modelo inicial es

$$ X \theta = y$$
Normalizamos la columnas de $X$ y llamamos a la nueva la matriz nueva $X_n$, la solucion al sistema cambia ahora es $\theta_n$, el nuevo sistema es

$$ X_n \theta_n = y \quad , \quad (5) $$ $y$ no cambia,

Pero nuevo problema es valido:

1. Hallamos $\theta_n$
2. Normalizamos los $X$ para datos nuevos con los parametros de normalizacion que fueron guardados en memoria
3. Revolvemos (5)

### Metodos de normalizacion:


1. Standarizacion: Se normalizan las columnas de $X$ para que tengan media 0 y desviacion estandard $\sigma=1$. Como se logra esto?

$$ X_j = \frac{X_j - \mu}{\sigma_j} \quad, \quad j=1,2, \cdots , n$$

donde
$$\mu_j = \frac{\sum_{i=1}^m x_{ij}}{m} $$
$$ \sigma_j = \sqrt{ \frac{\sum_{i=1}^m (x_{ij}-\mu)^2}{m}}  $$

2. Mean normalizacion (normalizacion por media)

$$X_j = \frac{X_j - \mu_j}{\max_i \{ x_{ij} \} - \min_i \{ x_{ij} \}} \quad , \quad i=1,2, \cdots, m$$

3. Min-max scaling (eslamiento por min/max)

$$X_j = \frac{X_j - \min_i \{ x_{ij} \}}{\max_i \{ x_{ij} \} - \min_i \{ x_{ij} \}} \quad , \quad i=1,2, \cdots, m$$

4. La normalizacion por la norma del vector

$$ X_j =  \frac{X_j}{\| X_j \|} $$
