## Problema de minimos cuadrados regularizado.

El problema (mas importante) del algebra lineal es la solucion de $Ax=b$.
Si $A$ es no singular, la solucion es posible:

$$ x = A^{-1} b$$.

Pero en general $A$ ni siquiera es cuadrada. Si $A$ no es cuadrada podemos tener dos tipos de problemas:

1. El sistema es **sobredeterminado** (overdetermined). Es decir hay  mas ecuaciones independientes que incognitas. La solucion no existe.
El espacio nulo es $\{0 \}$ pero el rango no cubre todo el espacio  ??

---



2. El sistema es **subdeterminado** (underdetermined). Existen mas incognitas que ecuaciones. El espacio nulo aca no es $\{ 0 \}$.
Hay un numero infinito de ecuaciones. 

El problema de **minimos cuadrados** viene "al rescate".
Ya que $Ax \ne b$ (para ningun $x$) o que $Ax = b$ para un numero infinito de valores podemos analizar el problema de la siguiente forma.

Si $Ax \ne b$ para ninguna valor de $x$, nimimicemos la diferencia

$$f(x) = \| A x - b \|^2 $$
Hallar el $\min_x f(x)$. 
Es claro que $f(x)$ es la funcion error, entre $Ax$ y $b$

Ahora, pero si hay un numero infinito de soluciones, podemos 
resolver el problema de minimos cuadrados y tomar la mas pequena de
todas las soluciones $\min \| x \|^2$. 

Como construir un problema que "mate los dos pajaros de un solo tiro" y
que sea estable.

Tikhonov (Ruso) postulo el siguiente problema:

$$ f(x) = \| Ax - b \|^2 + \| \Gamma x \|^2  \quad  (0)  $$
donde $\Gamma$ es una matriz positiva definida. Si usamos calculo matricial (la clase anterior) tomamos el gradidente de $f$ y lo igualamos a 0. $\nabla_x f(x) = 0$ (tarea). Entontramos la extension de las ecuaciones normales:

$$(A^* A + \Gamma^* \Gamma)x = A^* b $$ 
Estas son las ecuaciones normales extendidas (regularizadas) y siempre tiene solucion y unica.

En la proxima tarea les voy preguntar que muestren que si $\Gamma$ es positiva definida entonces $A^* A + \Gamma^* \Gamma$ tambien es positiva definidad y por lo tanto tiene inversa. Tambien les pregunto (primer punto) probar que toda matriz positiva definida tiene inversa
(recorderis: $B$ es positiva definida si todos sus autovalores son positivos, o de otra forma si $\forall x \in \mathbb{C}^n , x^T B x > 0 $)

De forma que la solucion:

$$ x = (A^*A + \Gamma^* \Gamma)^{-1} A^* b$$

Quien es $\Gamma$? Es comun en la literatura escoger un $\Gamma$ sencillo,
Por ejemplo $\Gamma = \sqrt{\lambda} I$, $\lambda > 0$. En este caso la solucion es:

$$ (A^*A + \lambda I)  x = A^* b  \quad  (1)  $$
Si $\lambda=0$ estamos hablando del **minimos cuadrados** de la clase anterior $A^* A x = A^* b$, la solucion es:

$$ x = (A^*A + \lambda I)^{-1} A^* b   \quad (2) $$.

A este problema que estamos atacando se le llama **regularizacion** y $\lambda$ es el **regularizador**. Hay ambiguedad. La pregunta es:   quien es $\lambda?$ .  A la ecuacion (1) se le llama en la literatura estadistica como **ridge regression** . Nosotros lo llamamos minimos cuadrados regularizados. 

[Levenberg-Marquardt](https://es.wikipedia.org/wiki/Algoritmo_de_Levenberg-Marquardt) trabajaron este problema y al $\lambda$ se conoce como el Levenberg-Marquard $\lambda$ o parametro de amortiguacion. 

Una nota: A los metodos de regularizacion que estamos viendo , tambien se le llaman: Tikhonov-Miller method o, Phillips-Twomey method. 


[Tikhonov Regularization](https://en.wikipedia.org/wiki/Tikhonov_regularization) . 

Seguimos con el misterioso $\lambda$. 

La funcion objetivo, correspondiente a la ecuacion (0) es

$$ J(x) = \| Ax - b \|^2 + \lambda \| x \|^2     $$

Este problema se puede ver como un problema de multiplicadores de Lagrange. Un problema de optimizacion con restricciones:

Es decir:

\begin{eqnarray} \min_x \frac12 \| Ax - b \|^2  \\
\text{sujeto a la restriccion} \| x \|^2 \le c  
\end{eqnarray}
Se puede probar que el numero que minimiza $f(x)$ es el mismo que minimiza $(1/2) f(x)$.

Estas son la ecuaciones normales. Podria pasar que $A^*A$ no tenga solucion. ## La Pseudo-Inversa o Inversa generalizada

El problema central es $Ax=b$. Sabemos que si $A$ no es cuadrada el problema no tiene solucion. Mejor, si $\det A=0$, si $A^{-1}$ no existe.
Quisieramos encontrar una matriz $A^+$ tal que

$x= A^+ b$ sea una solucion aceptable. 
* Que sea unica
* Que sea la "mejor"  ($\| Ax - b \|^2$) sea el "least" y $\|x \| $ el menor posible. 

Estudiemos casos:  (para resolver $Ax = b$.

* Si $b \in \mathcal{R}(A)$ la solucion existe. De aca salen dos casos:
    * Si $\mathcal{N}(A) = \{ 0 \} $. La solucion es unica y 
    $A^{-1}$ existe (con seguridad la matriz es cuadrada). La solucion es $x=A^{-1} b$, y $A^+ = A^{-1}$. 
    * Si $\mathcal{N}(A) \ne \{ 0 \}$. La solucion no es unica. Escoja la mas pequena de todas $\min \| x \|$.

## Como encontrar la pseudoinversa.
Keener (James P. Keener, referencia en las notas de clase) muestra (por lo menos) 4 metodos para hallar la pseudoinversa. Aca mostramos uno solo.

Una cosa buena de la pseudoinversa que **todas** las matrices $A$ tienen pseudoinversa. 

Usamos SVD. Recuerde que cualquier matriz $A$ (bonito de SVD).

$$ A = U \Sigma V^* $$ donde $\Sigma$ es diagonal con los valores singulares y $U$ , $V$ son matrices ortogonales con columnas autovectores de las matrices $A A^*$ y $A^* A$ respectivamente.

Recuerde que

$$\Sigma = \begin{pmatrix}
\Sigma_r & | & 0  \\
-- & -- & -- \\
0 & | & 0 
\end{pmatrix}  $$

Quisieramos $A^{-1} = V \Sigma^{-1} U^*$ 
El problema es la matriz de la mitad, los 0 no tienen inverso.

Definimos una matriz $\Sigma^+$, tal que los valores de la diagonal son los $\sigma_i^{-1}$ desde $i=1$ hasta $i=r$ y los demas 0,. 

$$\Sigma = \begin{pmatrix}
\Sigma_r^{-1}  & | & 0  \\
-- & -- & -- \\
0 & | & 0 
\end{pmatrix}  $$

Podemos escribir la matriz $\Sigma^+$ como

$$ \Sigma^+ = \begin{pmatrix}
\sigma_1^{-1}  & 0 & \cdots & \cdots & \cdots & \cdots & 0 \\
0 & \sigma_2^{-1} & \ddots & \ddots & \ddots  & \ddots  & \vdots \\
\vdots &   \ddots & \ddots & \ddots  & \ddots & \ddots & \vdots \\
\vdots & \ddots & \ddots & \sigma_r^{-1} & \ddots & \ddots & \vdots \\
\vdots &   \ddots & \ddots & \ddots  & \ddots & \ddots & \vdots \\
\vdots &   \ddots & \ddots & \ddots  & \ddots & \ddots & 0  \\
0 & \cdots & \cdots & \cdots & \cdots & \cdots & 0 
\end{pmatrix} $$

Dimensiones: 

$A$ es $m \times n$, $U$ es $m \times m$, $\Sigma$ es $m \times n$, 
$V$ es $n \times n$. Entonces $\Sigma^+ = n \times m$.
Entonces $A^+ = V \Sigma^+  U$ es:

$$ (n \times n) \times ( n \times m) \times (m \times m) $$
O sea que $A^+$ es $n \times m$. 
Si yo les pregunto que inviertan $A= (2 \quad 3)$, la solucion es


$$ A^+ = \begin{pmatrix} x \\ y  \end{pmatrix}$$

La pseudosolucion de $Ax=b$ en terminos de SVD es:

$$ x^+ = A^+ b = V \Sigma^+ U^* b  $$
