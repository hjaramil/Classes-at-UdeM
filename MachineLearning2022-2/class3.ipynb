1. Sea $A$ la matriz
$$ A = \begin{pmatrix} -5 & -3 \\ -6 & 2\end{pmatrix} $$

El polinomio caracteristico esta dado por

$$ p(\lambda) = \det(A - \lambda I) = \det
 \begin{pmatrix} -5-\lambda & -3 \\ -6 & 2-\lambda\end{pmatrix} =(-5 - \lambda)(2 - \lambda) -18
$$

Es decir
\begin{eqnarray}
p(\lambda) &=& -10+3\lambda + \lambda^2 - 18 \\
&=& \lambda^2 + 3 \lambda - 28 \\
&=& (\lambda + 7)(\lambda - 4)  
\end{eqnarray}

Es decir, los autovalores son la solucion $p(\lambda)=0$, o sea $\lambda=-7, \lambda=4$.

Como se hallan los autovectores? La idea es insertar estos valores en la ecuacion $(A - \lambda I) x = 0$.
Esto produce un sistema. Este sistema es singular, entonces hay infinidad de soluciones.

2. Encuentre los autovalores/vectores de la matriz

$$A= \begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix} $$.
Explique el significado de los autovalores/vectores usando un cuadrado de vertices $(1,1), (-1,1),(-1,-1),(1,-1)$ y un circulo $x^2 + y^2=1$.


**Solucion**:

\begin{eqnarray}
p(\lambda) ( A - \lambda I) &=& \det \begin{pmatrix}
1 - \lambda & 1/2 \\ 1/2 & 1 - \lambda
\end{pmatrix} \\
&=& (1-\lambda)^2 - \frac14 \\
&=& (1 - \lambda - 1/2) (1 - \lambda + 1/2) \\
&=& (1/2 - \lambda) (3/2- \lambda) 
\end{eqnarray}

Esto significa que los autovalores son $\lambda_1=3/2$ y $\lambda_2=1/2$


* Para $\lambda = 3/2$
Tomamos la primera ecuacion de $(A-\lambda I)x = 0$
para los autovectores.

$$ (1 -3/2) x + (1/2) y = 0$$.
$$ (-1/2)x + (1/2)y=0$$.

Una solucion facil (no normalizada) es
$$u = \begin{pmatrix} 1 \\ 1 \end{pmatrix} $$

* Para $\lambda = 1/2$.

$$ (1 -1/2) x + (1/2) y = 0$$.
$$ (1/2)x + (1/2)y=0 \quad (1) $$.

La solucion facil (no normalizada ) es

$$v = \begin{pmatrix} 1 \\ - 1 \end{pmatrix}  $$.

De la ecuacion (1) multiplicamos por 2

$$ x + y = 0$$.
Las solucion es $y = -x$.

Los autovectores (**todos**) son


\begin{eqnarray}

= \alpha \begin{pmatrix} 1 \\ -1 \end{pmatrix}
\end{eqnarray}
$\alpha \ne 0$

Por que,  si $u$ es autovector, entonces $\alpha u$ tambien es autovector $\alpha \ne 0$.

Sea $u$ autovector, o sea que $A u = \lambda u$ donde $\lambda$ es el autovalor (definicion).

$$ A (\alpha u) = \alpha A u = \alpha (\lambda u)= \lambda (\alpha u) $$

Teorema: Si a es autoadjunta (recuerde $A^*=A$), entonces los autovectores correspondientes a autovalores distintos son ortogonales y los autovalores son reales.

Prueba: Primero mostremos que los autovalores son reales.
Un ejercicio: si $\lambda = \overline{\lambda}$ entonces $\lambda$ es real. 

Calculamos la forma cuadratica $\langle A u_j, u_j \rangle $.

$$ \langle A u_j, u_j \rangle = \langle \lambda_j u_j, u_j \rangle = \overline{\lambda_j} \langle u_j, u_j \rangle = \overline{\lambda_j}  \| u_j \|^2 \quad  (2)  $$

De otro lado.

$$ \langle A u_j, u_j \rangle = (A u_j)^* u_j = u_j^* A^*  u_j =  u_j^* A u_j = u_j^* (A u_j) = \langle u_j, A u_j \rangle  \quad (3)  $$



Tenemos entonces de (2) y luego la (3)

$$ \overline{\lambda_j} \| u_j \|^2 = \langle A u_j, u_j \rangle = \langle u_j , A u_j \rangle = \langle u_j, \lambda_j u_j \rangle =  \lambda_j \langle u_j, u_j \rangle  = \lambda_j \| u_j \|^2  \quad (4)  $$

Como $\| u_j \| \ne 0$ dividimos por $\|u \|^2$ es ultima ecuacion y obtenemos $\overline{\lambda_j} = \lambda_j$. 

O sea que $\lambda_j$ es real.


Asumamos dos autovalores distintos $\lambda_1 \ne \lambda_2$. $\lambda_1$ correspsondde a $u$ y $\lambda_2$ corresponde a $v$. $u,v$ son autvectores.  Miremos la forma cuadratica.

$$\langle A u, v \rangle = \langle \lambda_1 u , v \rangle = \lambda_1 \langle u, v \rangle   \quad (5) $$

De otro ladod, como $A=A^*$,

$$\langle Au , v \rangle = \langle u, A v \rangle = \langle u, \lambda_2 v \rangle = \lambda_2 \langle u, v \rangle  \quad (6) $$
De las eqs 5, 6

$$ \lambda_1 \langle u , v \rangle = \lambda_2 \langle u, v \rangle $$ , pasando a restar pordemos escribir:

$$(\lambda_1 - \lambda_2 ) \langle u, v \rangle = 0 $$
y como $\lambda_1 \ne \lambda_2$ entonces $\langle u, v \rangle = 0$. 

Esto genera un resultado interesante:
Asuma que todos los autovalalores son diferentes entre si en el epacio $\mathbb{C}^n$. Entonces los autvectores construyen una base ortogonal en $\mathbb{C}^n$. Normalicemosla, es decir. La base seria

$$ B= \{ x_i  \ |  \langle x_i, x_j \rangle = \delta_{ij} \} \quad , i,j=1, \cdots , n$$

Tomemos formulas cuadraticas.

$$ x_i^* A x_j =  x_i^* (A x_j) = x_i^* \lambda_j x_j = \lambda_j x_i^* x_j = \lambda_j \delta_{ij}  $$






Esto en forma matricial se escribe como

$$ Q^* A Q = \Lambda  \quad (7) $$
Donde $Q$ es la matriz cuya columna $j$ es el autovector $x_j$ **normalizado** y $\Lambda$ es la matriz de autovalorese de $A$. 

Las matrices $Q$ son ortogonales (unitarias, ortonormales)
o sea que $Q^* Q = Q Q^*=I$. Entonces multiplicamos (7) a izquierda por $Q$ y a derecha por $Q^*$

$$ Q Q^* A Q Q^* = Q \Lambda Q^* $$

Es decir

$$A = Q \Lambda Q^*  \quad (8) $$
Esto es poderoso. Es el teorema espectral para autovaloress distintos. Veremos en la proxima clase el teorema correspondiente cuando los autovalores puedan ser repetidos. Es decir, basta con que los autovectores sean ortogonales para que la descomposicion espectral (8) sea un hecho.

Definicion de espacios invariantes.  Sea $W$ un espacio lineal (del algebra lineal, es el espacio generado por una base. $\langle x_1, x_2, \cdots,  x_n \rangle = \{ \sum \alpha_i x_i | \alpha_i \in \mathbb{C} \}  $)
decimos que $W$ es **invariante** de la matriz $A \in \mathbb{C}^{n \times n}$, sis $A W \subset W$. 

Ejemplos. 

* \{ 0 \} es un espacio lineal (trivial). Es invariante para cualquier matriz cuadrada. sea $x \in \{ 0 \}$,
entonces $Ax= A0 =0$, pero $0 \in \{ 0 \}$, es decirr
$Ax \in \{0 \}$ entonces $\{0\}$ por defincion es invariante. 
* $U= \mathbb{C}^n$ es invariante, pues $Ux \in \mathbb{C}^n$.  Pero estos son triviales y no son intersantes.
Busquemos ejemplos mas interesantes.

* El espacio nulo. Son $x$ tales que $Ax=0$. Probemos que $A \mathcal{N}_A \subset \mathcal{N}_A$
Prueba: sea $x \in A \mathcal{N}_A$, o sea que $x=0$. 
$x \in \mathcal{N}_A$. O sea que $\mathcal{N}_A$ es invariante.

* Si $x$ es autovector de $A$ entonces el genereado de $x$ es invariante. Prueba: $Ax = \lambda x$, y $\lambda x$ esta en el generado de $x$.

* Si $\{x_1, x_2, \cdots, x_k \}$ sosn autovectores de $A$ entonces sel generado de ellos es invariante de $A$.

$$ \langle x_1, x_2, \cdots, x_k \rangle = \sum_{i=1}^k \alpha_i x_i \quad , \quad \alpha_i \in \mathbb{C}. $$
