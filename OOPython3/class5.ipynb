# Object Oriented Theory: (Teoria de programacion orientada a objetos)

## Introduccion:
Motivacion: Por algun tiempo estamos ensenados a desarrollar software de forma de procedimientos y funcionan bien. Por que OO? Realmente para poder entender esto necesitamos codigos grandes. Para codigos pequenos no es necesario, es mas , estorba.

Vamos a listar una serie de puntos por los cuales el OO es importante.

* Modularidad: Como en la estacion espacial internacional, el software funciona muy bien con modulos independientes y conexiones entre ellos. Esto facilita el desarrollo y la depuracion de codigo.

* Simplificacion. La misma modularidad simplifica codigo, por que un modulo se puede usar en muchas ocaciones. Cuando es OO importante?
Bjarne Stroustrup (el que invento el C++) dice que la complejidad de un programa que no es OO crece de forma cuadratica como funcion del numero de lines del codigo. Si es OO crece de forma lineal. Esto no es un teorema pero tiene sentido. 

* Reusabilidad: Los objetos se pueden "instanciar" muchas veces sin repetir codigo. Por que el codigo esta todo determinado en la clase.

* Seguridad: Las clases permiten **esconder** codigo. Esto sirve para 
seguridad y seguridad. **safety** y **security** .
El desarrollo en OO provee seguridad contra robo. Proteccion de robo de codigo, como seguridad de que el codigo no vaya a sacar la mano.

Para resumir con respecto a OO:

* Es **confiable**: Por que se puede modificar codigo sin que afecte otras partes.

* Es **reusable**. Los objetos se pueden reusar en muchos contextos. El codigo se puede reusar mediante poliformismo, herencia etc.

* Es facil de  **mantener**. El soporte de un programa es el 80% el desarrollo el 20% (por decir). Los objetos son mas faciles de soportar por la forma modular.

* Es **extensible**: EL OO se puede extender con facilidad. Por ejemplo  agregar las patinetas a un sistema de transporte de una ciudad.

* Es **flexible**: Se puede desarrollar codigo sin necesidad de esperar que otros acaben lo que estan haciendo. Usted se concentra en su objeto (carro) y otros en otros objetos (moto, bilicleta, etc.)
