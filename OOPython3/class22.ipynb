# Abstraction:
## Abstract Clases: Clases abstractas


```python
from abc import ABC, abstractmethod # abc=libraria de clases abstractas
```

```python
# ejemplo con geometria 3D
from numpy import pi
class Solid(ABC):
    scale=5

    @abstractmethod
    def surfaceArea(self):
        pass

    def volumen(self):
        pass

# creamos clases que descienden de solido
class Cylinder(Solid):

    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def surfaceArea(self):
        return 2*pi*self.radius*self.height + 2*pi*self.radius**2

    def volume(self):
        return pi*self.radius**2*self.height

    def volumeScaled(self):
        return self.scale*self.volume()

class Sphere(Solid):
    def __init__(self, radius):
        self. radius=radius

    def surfaceArea(self):
        return 4*pi*self.radius**2

    def volumen(self):
        return (4/3.)*pi*self.radius**3

    def volumenScaled(self):
        return self.scale*self.volume()

def Cone(Solid):
    def __init__(self, radius, height):
        self.radius = radius
        self.hegiht = height

    def volume(self):
        return pi*self.radius**2*self.height/3
```
