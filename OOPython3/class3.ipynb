Underflow/Overflow
Notacion de distintos formatos de enteros en el computador

* entero (short or half) de 16 bits (sin signo) formato "h"
* entero (short or half) de 16 bits (con signo) formato "H"
* entero de 32 bits (sin signo) en formato "i"
* entero de 32 bits (con signo) en formato "I"
* numeros de punto flotante (IEEE).


Ejemplos

```python
# signed half
from array import array
a = array("h") # 16 bits signed

i= -2**(15)  # inferior
a.append(i)

s = 2**15 - 1
a.append(s)
print("rango de enteros medios o short", a)
```

```python
a.append(2**15)
```

```python
a.append(-2**15-1)
```

```python
# unsigned half
a= array("H")

# 16 bits 
i=0
a.append(i)

s=2**16-1
a.append(s)
print("range of unsigned halp is",a)
```

```python
a.append(2**16)
```

```python
a.append(-1)
```

```python
# integer 32 bits
a = array("i")
i=-2**(31) # inferior
a.append(i)

s = 2**31 -1
a.append(s)
print("range of unsigned integers",a)
```

```python
a.append(s+1)
```

```python
a.append(i-1)
```

```python
print(i-1)
```
