# Borrar atributos 4 combinaciones.
* De una clase:
    * variables de clase
    * variables de instancia
* De un objeto:
    * variables de clase
    * variables de instancia

Las cruzadas (clase con instancia , objeto con clase) no funcionan

```python
class Car():
    brand = "Toyota"

    def __init__(self, color="yellow"):
        self.color = color
        return


```

```python
myCar = Car()
vars(Car)
```

```python
# borremos variable de clase, desde la clase
delattr(Car, "brand")
vars(Car)
```
