# Analisis de 53 bits
Asuma el numero $a=0.2$.
Escribalo con 30 cifras decimales

# convertir 0.2 a IEEE con 30 digitos de precision.
\begin{eqnarray}
0.2 \times 2 &=& 0.4  \quad \text{parte entera es 0} \\
0.4 \times 2 &=& 0.8  \quad \text{parte entera es 0} \\
0.8 \times 2 &=& 1.6  \quad \text{parte entera es 1} \\
0.6 \times 2 &=& 1.2  \quad \text{parte entera es 1} 
\end{eqnarray}
A partir de este momento el numero es periodico.
Es decir, en precision infinita

$$ (0.2)_2 = 0011 \; 0011 \; \cdots  \; 0011 \cdots = \overline{0011}$$
Pensemos inicialmente en truncar este numero a 52 bits.


```python
a=0.2
print("a=%.30f"%a)
```

```python
period = "0011"
binary = ""
nbits = 52

# assuming 52 bits
for i in range(int(nbits/4)):
    binary += period

print(binary)
```

```python
# let us count bits
st  = str(binary)
print(nbits)
len(st) == nbits
```

# la primera cifra significativa es la tercera. Es decir, en notacion
# cientifica el numero es:
$001.1001100110011001100110011001100110011001100110011 \times 2^{-3}=(1+f) 2^{-3}$,


$$f = 1001100110011001100110011001100110011001100110011$$


```python
period = "1001"
fbin = ""
nbits = 52

# assuming 52 bits
for i in range(int(nbits/4)):
    fbin += period

print(fbin)
```

```python
# let us count bits
st  = str(fbin)
print(nbits)
len(st) == nbits
```

```python
# let us build the actual decimal using IEEE.

f = sum(2**(-i-1)*int(st[i]) for i in range(nbits))
print("mantisa", f)
decimal = (1+f)*2**(-3)  # c=1020
print("decimal %.30f"%decimal)
```

## Hmm.. me quede corto. Usemos 53 bits

```python
period = "1001"
binary = ""
nbits = 53

# assuming 53 bits
for i in range(int(nbits/4)): 
    binary += period

# since int(nbits)/4=52 we need to add the next bit by brute force
binary += "1"



print(binary)
```

```python
# let us count bits
st  = str(binary)
print(nbits)
len(st) == nbits
```

```python
# convert back to decimal
f = sum(2**(-i-1)*int(st[i]) for i in range(nbits))
print("f=%.30f"%f)

decimal = (1+f)*2**(-3)
print("decimal= %.30f"%decimal)

```
