El problema es que la clase ```Segment``` cambia el valor del atributo "point". Hay dos posibles soluciones:

* Cambiar la formula de distancia donde se conserva el punto. $$ d = \sqrt{(x_1-x_2)^2 + (y_1 - y_2)^2}$$
* Cada vez que se va a calcular la distancia se debe redefinir el punto. 

* La pregunta es: ¿existe una forma de evitar que la clase ```Segment``` cambie el valor original del punto?

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        return

    def norm(self):
        return ((self.x)**2 + (self.y)**2)**0.5

class Segment:
    def __init__(self, x1, y1, point):
        self.x1 = x1
        self.y1 = y1
        self.point = point

        # chequear que el punto es una instancia de Point
        if not isinstance(point, Point):
            print("signature arguments should be x,y, point. point de la clase Point")
            return "error"

        # getting the new point
        print("self.point.x, self.point.y", self.point.x, self.point.y)
        point.x = point.x - self.x1
        point.y = point.y - self.y1
        print("self.point.x, self.point.y", self.point.x, self.point.y)


        print("punto dentro de la funcion init (x,y)=", point.x, point.y)

        return

    def lengthSegment(self):
        # return self.point.norm()
        return self.point.norm()
````

````python
p1 = Point(5,6)
seg = Segment(3,4, p1) # estamos agregando el punto "point"
# seg = Segment(3,4,point) 

print(seg.lengthSegment())
````
