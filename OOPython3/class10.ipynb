# Overwrite the ```print()``` method. 



````python
class Point:
    def __init__(self, x, y, z = 0):
        self.x = x
        self.y = y
        self.z = z
        return

    # overwrite the function print
    def __str__(self):
        return f"X: {self.x}: Y: {self.y}, Z:{self.z}"

    
print(Point(-1,3))
print(Point(2,4,0))
print(Point(1,10,30))
````

# Class vairables in Python: Variables de Clase en Python.

Las variables de clase no se definen bajo ```__init__()``` las variables
cobijadas por ```__init__()``` se denominan de **instancia** (intance).

Para que son utiles? son utiles cuando **todos** los objetos tengan esa variable con el mismo valor. 

Por ejemplo $pi=3.14159....$ es una constante que en computacion es una "variable"

Las variables de clase no necesitan ```self``` .

Ejemplo

````python
class Car:
    brand = "Toyota"
    model = "Cololla"
    year = 2014

    def __init__(self, color="yellow"):
        self.color = color
        return

    def stop(self):
        print("please stop")
        return



````

````python
myCar = Car()
````

El operador **punto** ```.```
the dot operator

se puede usar el operator "." para accesar y hasta cambiar atributos de una clase

````python
myCar.brand
````

````python
# cuando el operador punto se usa con un metodo debe escribir () al final
myCar.stop
````

````python
myCar.stop()
````

El metodo ```del()``` es poderoso puede borrar cualquier obejto y sus atributos

````python
d = Car()
vars(d)
````

````python
class Car:
    brand = "Toyota"
    model = "Cololla"
    year = 2014

    def __init__(self, color="yellow"):
        self.color = color
        return

    def stop(self):
        print("please stop")
        return

    def muestreVariablesDeClase(self):
        print("brand=", self.brand)
        print("model=", self.model)
        print("year", self.year)
        return



````

````python
d = Car()
vars(d)
````

````python
d.muestreVariablesDeClase()
````

````python
# borremos el objeto d
del(d)
d.color
````

````python
c=Car()
c.color
````

````python
# borremos el atributo del objeto "c"
del(c.color)
````

````python
c.color
````

````python
e = Car()
e.color
````

````python
delattr(Car, "model")
````

````python
dir(Car)
````

````python
import inspect
inspect.getmembers(Car)
````
