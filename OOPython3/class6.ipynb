# Objetos y Clases

## Objetos. 
Un objeto en la realidad es "una silla", "un carro". Esto es concreto, 
la descripcion de la silla o el carro es abstracto. La descripcion se llama **clase**, la creacion, la fabricacion de la clase se llama **objeto**. En este sentido el objeto es una instancia de la clase. A esto se le llama **instanciacion**. En python todo son objetos.

El objeto (y la clase). Hay varios tipos de informacion:

* **datos**: atributos (adjetivos y sustantivos). 
* **metodos** que son verbos.

Dentro de los atributos existen categorias.

* publica: el atributo se ve para todo el mundo
* privada: solo autorizados (de la misma clase) pueden verlos.
* protegidos (protected). Solo lo ven los objetos de la misma clase y sus herederos.

Como se comunican los obejtos: A traves de mensajes y estos son los metodos o verbos que se describen arriba y de los cuales mucho vamos a decir.

## Clases:
* Una clase es el **blueprint** de un objeto.  Es la descripcion del objeto, es el mapa del objeto, es la patente del objeto. Es la abstraccion del objeto.

* La clase tiene atributos y metodos.   

* Decimos que el objeto es una instanciacion de la clase o una instancia de la clase.



## Acerca de los nombres: 

* Las clases, se aconseja, que comiencen con mayuscula y que cada palabra  dentro del nombre comience con mayuscula. Por ejemplo
ReinoAnimal*  , *AnimalKindom*. A este tipo de notacion se le llama
**CamelCase** .

* Los objetos se notan comenzando con minuscula y cada palabra nueva con mayuscula. Por ejemplo *gireDerecha*, *oprimaElAcelerador*. 
A este tipo de denominacion se le llama "dromery case"

Se aconseja, cundo quiera ser desarrollador serio (profesional) de Python seguir el estandard en:
[PEP-8 Guide](https://peps.python.org/pep-0008/). 

Se usa tambien separar palabras con guiones bajos. 
gire_derecha, gire_izquierda. Pero yo los uso poco.

## Herencia: Inheritance ( relacion "is a").
La herencia permite reciclar codigo de forma "impresionante".

Por ejemplo, la clase "AnimalKindom" tiene una cantidad de atributos impresionante, que por supuesto tienen todos animales. 
Cuando definimos una nueva clase, "Mammal" (Mamifero) podemos, automaticamente, heredar todos los atributos y metodos de la clase
"AnimalKindom".  Y mas alla, la clase "Dog" herereda todo del "Mammal".
Con esto se evita re-escribir codigo.

Teminologica: La clase "child" hereda de la clase "parent".
A la clase "parent" se le llama "base class" (clase base) y 
a la clase "child" se le llama "derived class".
Tambien se usa la terminologia "subclass" (para child) y 
superclass bara parent

* parent $\leftrightarrow$ superclass $\leftrightarrow$ baseclass
* child $\leftrightarrow$ subclass $\leftrightarrow$ derived class.

Como trabaja la herencia

$$ a( x + y + z + w) = ax + ay + az + aw $$
