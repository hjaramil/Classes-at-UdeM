Python usa "DocStrings" que toman la documentacion directamente de los comentarios.

Ejemplo

```python
class Car:

    ''' La clase "Car" describe cosas importantes de los carros'''

    def __init__(self, brand="Nissan", model="Sentra", year=2015, engineTemperature=100):
        ''' inicializacion muestra los parametros por defecto y sus argumentos
         otro comentario 
         '''

        self.brand = brand
        self.model = model
        self.year = year
        self.engineTemperature = engineTemperature

        return



    def temperaturePlusOne(self):
        ''' suma un grado de temperatura a la actual
        otro comentario en temperaturePlusOne '''
        return self.engineTemperature + 1

    def getTemp(self):
        " obtiene la temperatura"
        return self.engineTemperature

    def setTemp(self, newtemp):
        " asigna una nueva temperatura"
        self.engineTemperature = newtemp
        return

test = Car()
```

```python
print("temperatura + 1 igual a ", test.temperaturePlusOne())
```

```python
print("brand", test.brand)
print("model", test.model)
print("year", test.year)
```

```python
test.setTemp(110)
print("temperatura igual a", test.getTemp())
```
