# Association (Asociación)
## Composition (Composición)
Es parte de. Por ejemplo la palanca de cambios es parte de un carro. Explicamos composicion con un ejemplo

````python
# la clase Point (punto)
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        return

    def norm(self):
        return (( self.x)**2 + (self.y)**2)**0.5

class Segment:  # segmento
    
    def __init__(self, x1, y1, x, y):
        self.x1 = x1
        self.y1 = y1

        # corrimiento del origen
        self.x = x - x1
        self.y = y - y1

        self.point=Point(self.x, self.y) # composicion
        return

    def lengthSegment(self):
        return self.point.norm()


seg=Segment(3,4,5,6)
print(seg.lengthSegment())



````

````python
# verifiquemos
import numpy as np
p1 = np.array([3,4])
p2 = np.array([5,6])
dist = np.linalg.norm(p1-p2)
dist
````

    ```encuentraNota(id)```

* El metodo ```imprimaNota()``` imprime **todas** las nota. Puede usar el metodo ```vars()``` para esto. La firma

    ```imprimaNota()```
