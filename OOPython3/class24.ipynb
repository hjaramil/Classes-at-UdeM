### Scope: Alcance
scope es el alacance que tiene una variable. ESta puede ser

* En tiempo. Tiempo en el cual la variable esta "viva", activada.
* En espacio el conjunto de lineas de codigo donde la variable tiene validez

Otras categorias son

* local: solo tiene valor dentro de un rango de lineas de codigo limitado. Por ejemplo dentro de una funcion
* global: Durante toda la ejecucion del programa, en cualquier parte

````python
a=5  # a vale 5
a=6 # aca fallece el valor que llevaba a "a" de 5.
print(a)
def areaOfCircle(r):
    pi = 3.14159
    print("variables locales son\n", locals()) # locals() es un metodo global de python
    return pi*r**2

print(areaOfCircle(1))
````

### Global Scope


````python
def areaOfCircle(r):
    global pi  # no es remendado
    pi = 3.14159
    return pi*r**2

print(areaOfCircle(1))
````

````python
globals()
````

````python
# parece ser que las variables locales y globales son iguales
locals()==globals()
````

## Exceptions : excepciones

Antes de estudiar las excepciones debemos entender el tipo de errores que ocurren en un sitema. 

* Errores de compilcacion: Son los mas faciles de arreglar (sintaxis), olvidar una coma, un :, indentacion, 

* Errores logicos: Pueden ser dificiles. Por ejemplo

$$ a/b+c  = a/(b+c)$$



* Run time error: aparece al momento de correr el programa. Puede ser un NaN (Not a number), raiz de un numero negativo (en el campo real). 

````python
a=2
b=3
c=4
print(a/b+c)
print(a/(b+c))
````

````python
a/b+c == a/(b+c)
````

De acuerdo al estado de corrido del programa exiten dos tipos de errors:

* Un "fatal error". El programa para de correr
* Un "no fatal error" . El programa sigue corriendo aun despues del error.

Otro nivel de clasificacion:

* Falla de Software (Hardware failure)
* System error (por ejemplo accesar archivos sin autorizacion)
* Software exceptions.   Por ejemplo un programa donde las densidades negativas se rechazan.
