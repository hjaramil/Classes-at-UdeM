## islice: 
Tiene el mismo papel que tiene en NumPy, arange, linspace
Syntaxis: 
```itertools.islice(iterable, start, stop, step)```
donde
```iterable``` es un iterador
```start``` el comienzo
```stop``` es el final
```step``` es el brinco 


````python

result = itertools.islice(inList, 1, 15, 2)
print(list(result))
````

## Compress
la syntaxis:

```istertools.compress(data, selectors)```
Este actua como un filtro seleccionando elementos de unos datos.

Se parece a la siguiente formula

```
def compress(data, selectors):
    return (d for d, s in zip(data, selectors) if s)

Ejemplo
```compress('ABCDEF', [1,0,1,0,1,1]) --> A,C,E,F ```






```

````python
selectors = [1, 0, 1, 0, 1, 1]
letters='ABCDEF'
result = itertools.compress(letters, selectors)

result
````

## filterfalse().
La sintaxis es 
```itertools.filterfalse(predicate, iterable)```

Lo que hace es mostrar aquellos que sean "F" en el iterable.

````python
L= range(11)
print(list(L))
````

````python
def le_6(n):
    return n<=6

# filtro clasico
result = filter(le_6, L)
print(list(result))
````

````python
# veamos ahora el filtro falso
result = itertools.filterfalse(le_6, L)
print(list(result))
````

## dropwhile()
La sintaxis:
```itertools.dropwhile(predicate, iterable)```

Esta funcion elimina (drop) los elementos del iterable siempre que el predicado sea cierto. Luego retorna el **resto** de los elementos. El predicado es una funcion, esta puede ser una **lambda function**



````python
L=[1,4,6,4,1]
result = itertools.dropwhile(lambda x: x<5, L )
print(list(result))
````

## takewhile()
Sintaxis:
itertools.takewhile(predicate, iterable)
Este es realmente el complemento del ```dropwhile()```. Retorna los elementos del predicado hasta que el predicado de haga falso. Una vez el predicado se hace falso **para**.



````python
L=[1,4,6,4,1]
result = itertools.takewhile(lambda x: x<5, L)
print(list(result))
````
