## Clases y Objetos
Que es una clase? Una clase es un mapa que describe objetos.
Un objeto es algo que se crea de ese mapa. Una clase genera tantos
objetos como queramos. El objeto es "concreto" . Es una "concrecion" de la clase, en ingles "instantiation" (instanciacion) de la clase. 
La clase es una "abstraccion" del objeto. 


La clase tiene

* Atributos: Sustantivos y adjetivos. Se llaman **estados** de la clase
* Metodos: Verbos (metodos). **comportamientos** (behavior).

## Para que clases?

Ilustramos con un ejemplo cuando las clases estorban


````python
def sqr(x):
    return x**2

print(sqr(3))
print(sqr(4))
````

````python
class Square:
    side = 2  # el parametro por defecto

    def sqr(self):
        return self.side**2   # encuentra el area del cuadrado

# crear objeto
mySqr =Square()
mySqr.side=3 # cambio el parametro por defecto
a = mySqr.sqr()
print(a)
mySqr.side = 4
print(mySqr.sqr())
````

Que es **self** ?
El self es la memoria del objeto.
Verifiquemos esto



````python
class Square:
    side = 2

    def sqr(self):
        print("memoria del objeto ", id(self))
        print()
        return self.side**2

mySqr = Square()
mySqr.sqr()
print(id(mySqr))
````

````python
class Square:
    side = 2

    def sqr(self, a):
        print("memoria del objeto ", id(self))
        print()
        print(a)
        return self.side**2


a=5
mySqr = Square()
mySqr.sqr(a)
print(id(mySqr))
````

## metodo   \_\_init\_\_

Es util, cuando se crea una clase, incluir parametros que la clase va a necesitar siempre y que no hay re-asignarlos. 

Por ejemplo un carro siempre tiene 4 llantas y no queremos re-asignar este valor. Todos esos parametros, por defecto, que no queremos re-asignar van en el metodo \_\_init\_\_ . 

Comparando con C++ y Java, alli a este metodo lo llaman el **constructor**. Ahora bien en C++, por cada constructor hay
un **destructor**. Python (como Java) tienen garbage collection y ellos se encargan de destruir los objetos cuando ya no se necesitan. 

Python tambien tiene un destructor (no se usa mucho) que se llama
**del()**. Lo vamos a estudiar mas adelante. 

Visitando mas el interior de Python existe un metodo llamado **new()** que genera objetos.


### new()
Hay algo en la ciencia de la programacion que se llama **signature**.
El signature, se refiere a un metodo, tiene tres elemetos

* El nombre del metodo
* Los argumentos (entre parentesis)
* El tipo de retorno (para C, C++, Java, etc. No python

En el caso de new(), el "signature" es

```  object.__new__(class, arguments) ```





Veamos ejemplos del uso del **new()**




````python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self. model = model
        self.year = year
        return

myCar = Car("Toyota", "Corolla", 2010)
````

````python
myCar.__dict__
````

````python
# como usar el new() una nueva herramienta
class Car:
    def __new__(cls, brand, model, year):
        print(f'Creating a new {cls.__name__} object...')
        obj = object.__new__(cls)
        print(id(obj))
        return obj


    def __init__(self, brand, model, year):
        print(f"inicializador del objeto Car")
        self.brand = brand
        self.model = model
        self.year = year
        return

myCar = Car('Toyota', 'Corolla', 2010)
myCar2 = Car("Nissan", "Sentra", 2020)
print(id(myCar))
````
