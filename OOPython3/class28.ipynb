# Containers: contenedores:

Hay varios tipos de contenedores:

* Built in types:
    * list: listas, mutables, tienen indice, y orden
    * tuple: tuplas, inmutables, tiene orden y indices
    * dict: Diccionarios. Tienen key (llave) y valor. Los valores se accesan a traves de las llaves
    * set: conjuntos. No tienen indices, no hay repidos, creo que no tienen orden

* Contenedores de del modulo ```collection```. Hablaremos poco de esto.

Como sabemos si un objeto es contenedor o no:

* Haga ```dir``` en el objeto , si aparece el metodo ```__contains__```  entondes es un contenedor, sino, no.

* Otra forma es importando el modulo ```collections```. Y preguntar con la instruccion, si ```a``` es el objeto que estamos investigando

```isintance(a, collections.abc.Container)``` esta es buleana. Si retorna ```True``` es un contenedor.

## Contenedores Built-in:

* lists: listas. Una ```list``` es una colleccion de objetos, no necesariamente del mismo tipo. Se puede ordenar, si son del mismo tipo, es indizada. Pueden cambiar (es mutable). Se puede iterar por ellas.

La sintaxis es

`````python
myList = [1,5, 'hello', 'abc', 2, 3] # no es homogenea
`````

### Slicing: Sacar tajadas.

La sintaxis de "slicing" es:

```[start: end: increment]```, el ```increment``` puede ser negativo y si este no esta dado, se asume por defecto 1.
```start, end, increment``` son indices

Si el ```start``` no aparece se asume 0.
Si el ```end``` no aparce se asume el ultimo
Para accesar el elemento con indicie ```i``` de la lista ````a``` se usa ```a[i]```.


`````python
myList[1]
`````

`````python
# el elemento de indice 4 (2+2) no se muestra por que el "slicing" se abierto a derecha
myList[2:4:2]
`````

`````python
myList[4:1:-1]
`````

`````python
myList = []
myList == list()
`````

`````python
myList = list(range(10))
myList[-3]
`````

`````python
# muchos atributos en las listas
print(dir(myList))
`````

Algunos de los atributos importantes de las listas:

* append: La lista puede agregar elementos nuevos al final.
* insert: Inserta un elemento de la lista en cualquier parte.
* remove: remueve un elemnto (usando un indice) de la lista
* pop: Remueve el ultimo de la lista
* sort: ordenar de abajo hacia arriba o al contrario
* reverse: es reversar la lista

Ejemplos:




`````python
myList = list([3,2])  # o simplemente myList=[3,2]
print("myList is", myList)
myList.append('hi')
print("appending `hi` to my list yields", myList)
myList.insert(10, 'hello')  # como 10 es mas grande que todos los indices lo pone al final
print("inserting  `helo` al final de la lista", myList)
myList.remove('hi')
print("luego de remover 'hi' tenemos",myList)
# sobre-escribir
myList = [3,5,-1,5,0]
print("my nueva lista, luego de cambiarla",myList)
myList.sort()
print("my lista luego de ordenarla",myList)
myList.reverse()
print("mi lista en reversa", myList)
`````

## List comprehension
La sintaxis:

newList = [ expression ```for``` item ```in``` iterable ```if``` condition==```True```]

* expression: Es una funcion del ```item```
* item : Es la variable independiente
* iterable: es una lista, un rango, una tupla, un container
* condition: boolean

Por ejemplo para determinar si un numero is impar se puede
hacer de dos formas:
* ```i%2==1```
* ```i%2```



`````python
myList = [i**2 for i in range(10) if i%2 == 1]
myList
`````

`````python
myList = [i**2 for i in range(10) if i%2]
myList
`````

## Tuplas: 
Es una coleccion de objetos, no son mutables, es mas rapida que la lista.

* Es inmutable. No se pueden cambiar. Esto es bueno por seguridad
* Es ordenada, es indizada
* Puede ser seccionada (slice)
* Se pueden usar para diccionarioes, o como retorno de funciones que retornan varios valores
* Se notan con parentesis, como por ejemplo ```a=(1,3,5)```
* Los parentesis no son obligatorios. ```a=1,3,5```

`````python
a=(1,3,4,5,5,-1,2)
print(a[2])
`````

`````python
a[1:7:2]
`````

`````python
dir(a)
`````

`````python
print(len(a))
`````

`````python
# tamano
a.__sizeof__() # dunder, tamano en bytes
`````

`````python
a.index(5)
`````

`````python
# tuplas con clases
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        return

student = Student("Robert", 21)

# uso de tupla en clases
name, age = student.name, student.age
print(f"name={name}, age={age}")
`````

`````python
# otro ejemplo
myStr = 'Robert-31-student'
name, age, occupation = myStr.split('-')
print(f"name={name}, age={age}, occupatoin={occupation}")
`````

## Diccionarioes:
La sintaxis, es un conjunto con cada elemento con sintaxis 
```key:value```.

* Es mutable
* Muy buenos para busquedas, mediante "key"s.
Los elementos se accesan por la llave ```key```.

`````python
myDic = {"a":1, "b":2}
myDic["a"]
`````

`````python
# cast  a tuple into a dictionary
t = ( (1,'a'), (2, 'b'))
{y:x for x,y in t}
`````
