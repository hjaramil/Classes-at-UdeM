# Formatos numericos en la computacion: IEEE (no tildes é, ñ, no enhes)

 Numeros enteros en la computacion. Para la representacion de un numero se necesitan **simbolos** y **cifras**. Las cifras son los lugares reservados para los simbolos. 

 Cual es la menor cantidad de simbolos que se usan para representar nuestros numeros enteros.

 0? no por que esto significa que no hay simbolos y nada representar.

 1? No por que si un simbolo es, por decir, 1, necesitariamos un millon de cifras para representar el numero 1 millon. Y es demasiado. Pero peor aun si ese simbolo fuese 0, entonces como representamos el 1? 

 2? Si es un numero valido. De hecho los computadores de hoy trabajan en sistema binario. Con dos simbolos "1", "0". A estos dos numeros se les llama "bits".


 Se pueden usar cualquier numero de simbolos mayor a 2,
 3,4, etc. Los mas usados son 

 10, 8, 16  (decimal, octal, hexadecimal). 

 Como se interpretan estos simbolos para crear numeros.

 Pregunta de combinatoria. Cuantos numeros enteros se pueden generar con dos simbolos y dos cifras. La respuesta es 4.

 $$aa, ab, ba, bb $$
 Si $a=0, b=1 00, 01, 10, 11$. Se puden convertir a decimales usando la regla mostrada en el ejemplo (*)


$$ (10)_2 = 0 \times 2^0 + 1 \times 2^1 = (2)_{10} = 2$$ 

Generalicemos:

El siguiente numero binario (dos simbolos) de $n+1$ cifras
se representa en formato binario y decimal como sigue
$$ (b_n b_{n-1} b_{n-2} \cdots b_0)_2 = b_0 \times 2^0 + b_1 \times 2^1 + \cdots b_n \times 2^n$$

Ejemplo

$$ (011011)_2 =  1 \times 2^0 + 1 \times 2^1 + 0 \times 2^2 + 1 \times 2^3 + 1 \times 2^4 = 27$$



```python
0b011011
```

### Combinatoria.

Tenemos 3 cifras y 2 simbolos. Cuantos numero podemos simular.

Podemos simular 8 numeros. En decimal 0, 1, 2, 3, 4, 5, 6, 7.


Tenemos 16 bits, (binario). El numero total es $2^{16}$.
Con esto podemos generar, sin signo
$$ 0, 1, \cdots 2^{15}-1$$

```python
2**(15)-1
```

Pensemos en el signo. El primero bit puede represntar el signo. Si el primer bit es "1" el signo es negativo , si el primer bit es "0" el signo es positivo.

Ej:


$$ (1111)_2 = -(1 \times 2^0 + 1 \times 2^1 + 1 \times 2^2)=-7  $$

Si tenemos 16 bits el rango de valores con signo es

$$ -2^{15} \le 0 \le 2^{15}-1 $$

Otro ejemplo. En hexadecimal, tenemos 3 cifras. Cuantos numeros, sin signo, se pueden generar. 

$$ 16 \times 16 \times 16 = 16^3 = (2^4)^3 = 2^{12} = 2^10 \times 4 = 4096.$$

### Si yo doy un numero en base 10 como se representa ese numero en base binaria.

Ejemplo

$$ 10 = 5 \times 2 \quad , \quad 5 = 2(2) + 1 \quad, \quad 2 = 2(1) + 0 $$ 

combinando esto tenemos:\


$$ 10 = 5 \times + 0 = (2 (2) + 1) \times 2 + 0 = 2 + 2^3 = 0 \times 2^0 + 0 \times 2^2 + 1 \times 2^3 = (1010)_b  $$
