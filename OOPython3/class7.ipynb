Para explicar el valor de la "basurita" al final debemos entender como trabaje el sistema internamente.

Inicialmente les ensene a convertir de $0.1$ a binario con punto flotante.


\begin{eqnarray}
0.2 \times 2 &=& 0.4 \quad \text{parte entera es 0} \\
0.4 \times 2 &=& 0.8 \quad \text{parte entera es 0} \\
0.8 \times 2 &=& 1.6 \quad \text{parte entera es 1} \\
0.6 \times 2 &=& 1.2 \quad \text{parte entera es 1} \\
\end{eqnarray}
En este momento "cerramos el circulo" y regresamoa s 0.2.

Por lo que $0.2$ en formato binario con punto flotante es:


$$0011 \; 0011 \; \cdots  = \overline{0011} $$

Recordemos la representacion cientifica.
La mantisa pareciera ser

$$.001100110011  \cdots $$
$$1.1001 \; 1001 \; \cdots \times 2^{-3} $$
Tenemos lo siguiente


$$0.2 \approx (1 + f) \times 2^{-3}$$

```python
# construyamos la mantisa 1001 1001 1001
period = "1001"
fbin = ""
nbits  = 52

for i in range(int(nbits/4)):
    fbin  += period

print(fbin)
```

```python
st = str(fbin)
print(nbits)
len(st) == nbits
```

la mantisa se calcula con formula

$f = \sum_{i=1}^{nbits} 2^{-i} $

```python
# conversion a decimal
# la mantisa
f =  sum( 2**(-i-1)*int(st[i]) for i in range(nbits))
decimal = (1+f)*2**(-3)


print("decimal, %.30f"%decimal)
```

```python
# lo mismo pero con 53 bits en vez 52.
# construyamos la mantisa 1001 1001 1001
period = "1001"
fbin = ""
nbits  = 53

for i in range(int(nbits/4)):
    fbin  += period

# agrego el ultimo a la fuerza
fbin +="1"
print(fbin)
```

## Polimorfismo: Varias formas. Los objetos pueden tomar varias formas de acuerdo al contexto.

Vamos a estudiar, principalmente, dos tipos de polimorfismo.

* A traves de la herencia. Los objetos que heredan de la clase padre, podrian compartir un metodo pero ese metodo funciona distinto para cada objeto segun su clase. Piense en la clase **Animal**. Piense en el metodo **move**.  Este metodo puede acarrear muchos atributos (desde, hasta, trayectoria, velocidad, aceleracion, etc), sin embargo cuando
descendemos al hijo "serpiente" o "ave" , estos no comparten la misma forma de desplazarse, y alli se "polimorfean", es decir que cada uno se mueve de forma distinta a su manera.  

* Method overloaling: Sobrecarga de metodo. Asuma que un metodo requiere dos parametros, por ejemplo $sume(a,b)$ retorna $a+b$. Pero luego
aparecen 3 parametros $sume(a,b,c)$ y retorna $a+b+c$. 

* Operator overloading: Por ejmplo sean $a,b$ enteros $a+b$ es suma de enteros. Si $A,B$ son matrices $A+B$ es suma de matrices

## Scope (Alance).

Alanace: Es el intervalo de tiempo durante el cual una variable aparece habilitada en el sistema. Existen distintos alcances:

* Local scope: Una variable esta definida dentro de una funcion. Una vez la funcion sese de correr, la variable desaparece. Vamos a ver ejemplos en Python mas adelante.

* Global Scope: una variable declarada como global puede permanecer activa en el sistema (en memoria) aun cuando esta por fuera de una funcion declarada. No es recomendable para una programa tener muchas variables globales. Esto dificulta el "debugging" y el mantenmiento del programa. Vamos ejemplos en Python.

* Buil-in scope. Son variables que desde que el sistema se dispara hasta que termina tienen vida.
