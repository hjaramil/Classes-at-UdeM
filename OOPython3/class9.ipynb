# Encapsulation: Encapsulaci√≥n
La encapsulacion, que consiste en poner datos y metodos juntos en una sola capsula (clase) tiene las siguientes caracteristicas

* Information hidding (ocultar informacion). Esta informacion se oculta a objetos de otras clases y hasta de la misma. Tiene varios objetivos.
    * Proteger sus codigo o datos de usted mismo.
    * Proteger contra otros. Un ejemplo sencillo. Piense en un carro.
las direccionales se usan para indicarle a los peatones u otros conductores que usted va a girar a la derecha. La temperatura de su motor solo interesa a usted. El que otros conductores puedan leer la temperatura de su motor es un riesgo por leer informacion que es distractora. 

* El reuso. Reciclaje. Una vez se definine claramente una clase, esta se puede usar una y otra vez para crear muchos objetos sin tener que volver a definir sus caracteristicas.

* Modularidad (modularity).  Se construyen modulos que se pueden empatar con otros para crear sistemas complejos y faciles de visualizar y operar. Piense en la estacion interacional. 

* Ocultar complejidad. Esto usa mediante las interfaces con las cuales se evita mirar el codigo en su interior.

Yo digo, que "encapsulation" se puede llamar tambien "enclassulation" por que se construyen classes a partir de propiedades y comportamientos.

Vamos a ilustrar con ejemplos la construcion de clases donde se "oculta" informacion. 

Pensemos en el ejemplo del carro donde queremos ocultar la temperatura por que no es del interes de objetos externos a la clase.

Una forma es declarar la temperatura como una variable **privada** y estos se hace con un guion bajo antes de la palabra. Por ejemplo ```_engineTemperature``` . Sin embargo esto no oculta el dato.
Esto es una convencion para que las personas que usen el codigo no toquen esta variable. Sin embargo se puede ocultar con un doble guion bajo. ```__engineTemperature''' (nota: no hay nada oculto en Python).


Hay dos tipos de metodos importantes para accesar datos en OO

* **setters**:  son  metodos que pueden cambiar un valor.
* **getters**: pueden obtener un valor. Vamos a ver estas cosas con ejemplos.








````python
class Car:
    def __init__(self, brand, model, year, engineTemperature):
        self.brand = brand
        self.model = model
        self.year = year
        self.__engineTemperature = engineTemperature # aparentemente protegida
        return

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1

test = Car("Toyota", "corolla", 2015, 79)
test.temperaturePlusOne()

print(test.brand)
print(test.model)
print(test.year)




````

````python
print(test.__engineTemperature)
````

````python
## Ejemplo con setters and getters
class Car:
    def __init__(self, brand, model, year, engineTemperature):
        self.brand = brand
        self.model = model
        self.year = year
        self.__engineTemperature = engineTemperature # aparentemente protegida
        return

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1

    def getTemp(self):
        return self.__engineTemperature

    def setTemp(self, newtemp):
        self.__engineTemperature = newtemp
        return





test = Car("Toyota", "corolla", 2015, 79)
print("temperature +1=", test.temperaturePlusOne())

print(test.brand)
print(test.model)
print(test.year)

print("temperature returned from the object before change", test.getTemp())

## cambio de temperatura
test.setTemp(110)

print("temperature returned from the object after change", test.getTemp())








````

## Como manejar parametros por defecto: by default


````python
# todos por defecto
class Car:
    def __init__(self, brand="Nissan", model="Sentra", year=2015, engineTemperature=100):
        self.brand = brand
        self.model = model
        self.year = year
        self.__engineTemperature = engineTemperature # aparentemente protegida
        return

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1

    def getTemp(self):
        return self.__engineTemperature

    def setTemp(self, newtemp):
        self.__engineTemperature = newtemp
        return





test = Car()
print("temperature +1=", test.temperaturePlusOne())

print(test.brand)
print(test.model)
print(test.year)

print("temperature returned from the object before change", test.getTemp())

````
