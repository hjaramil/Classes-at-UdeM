# Inheritance: Herencia
Recordemos que cualquier clase en Python hereda de la superclase (Adan) que ```object``` . Esto lo ilustramos enseguida.

````python
objectMembers =dir(object)
print(objectMembers)
````

````python
print(object)
````

````python
a=2
intMembers = dir(a)
print(intMembers)
````

````python
dir(int)
````

````python
# para ver que de comun tienen dos conjuntos de atributos
# les hacemos "casting" (convertir) a conjuntos y 
# usamos las propiedades de conjuntos
objSet = set(objectMembers)  # convierte lista a conjunto
intSet = set(intMembers)

interSection = objSet.intersection(intSet)
print(interSection)
````

````python
# vamosa verificar que la interseccion es toda de la clase ```object```
# primero veamos el numero de miembros en cada caso
print(len(objSet))
print(len(interSection))
````

````python
objSet == interSection
````

Claramente todos los miembros de la clase ```object``` forman parte de cualquier otra clase.
Son **heredados**. Por ejemplo, 

```__class__```, ```__delattr__```, etc
Hay otros que no forman parte de la clase ```object``` pero si de la clase ```int```. 

## Herencia simple. La forma mas simple de herencia.
Existen aca dos clases. La clase padre
(parent=superclass=base class) la clase hijo.
(child=subclass=derived class).

Muestro un ejemplo matematico sencillo que motive la definicion de herencia.


$$ a(x+y+z)=ax + ay + a z $$

imagen.png

````python
# Ejemplo
class Person:
    def __init__(self, firstName, lastName, birthDay):
        self.firstName = firstName
        self.lastName = lastName
        self.birthDay = birthDay
        return

# probemosla
myPerson = Person("Bill", "Star", "06/21/84")
myPerson.__dict__
````

````python
# la forma mas simple de herencia entre dos clases es compartir
# el metodo __init__()
class Student0(Person):
    def pleaseSit(self):
        print("Por favor tome asiento")
        return

p1 = Person("George", "Bleistein", "13/08/60")
s1 = Student0("Mary", "Ann", "20/10/76")
print(vars(p1))
print(vars(s1))
````

````python
s1.pleaseSit()
````

````python
p1.pleaseSit()
````

````python
# podemos tomar solo algunos atributos de la clase padre y definir otros
class Student(Person):

    def __init__(self, firstName, lastName, birthDay, grade):
        super().__init__(firstName, lastName, birthDay) # hereda de la superclase

        self.grade = grade
````

````python
myStudent = Student("Jorge", "Agudelo", "06/21/1992", 11)
myStudent.__dict__
````
