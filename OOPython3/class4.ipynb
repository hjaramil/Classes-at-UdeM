## Paradigma imperativo.
Es el mas y el que usamos en el curso. Hay un flujo de control extricto. Ejemplos C, C++, Fortran, Basic, Pascal, Cobol, Python, Java. 

Del imperativo tenemos tres ramas

* Programacion estructurada
* Programacion de procedimientos (procedural programming)
* Programacion orientada a objetos OOP: Object Oriented Programming (este curso)

### Paradigma de procedimientos: 
Encima de la programacion estructurada se definen **subrutinas** (hoy **funciones**). Facilitan el trabajo.
Ejemplo en python


```python
# imprima los cuadrados del 1 al 10.
for i in range(1,11):
    print(3 + i + 2*i**2 + i**3 + 4*i**4)
```

```python
# mas elegande es

def s(i):
    return 3 + i + 2*i**2 + i**3 + 4*i**4

for i in range(1,11):
    print(s(i))
```

### Programacion Orientada a objetos:OOP (Object Oriented Programming).

Las funciones son verbos. En OOP las funciones se llaman metodos. Los objetos son cosas de la vida diara. Por un ejemplo un carro. Un carro tiene

* Sustantivos: puerta, llanta, frenos, etc
* Adjetivos: Verde, pesado, grande
* Verbos: acelera, frena, a la derecha, etc.

En OOP:

* atributos: sustantivos y adjetivos
* metodos: son los verbos.

Un clase, es un mapa de los objetos que se van a generar. Es "blue print" (la patente). 


No hablamos mas de programacion orientada a objetos por que ese es basicamente este curso. Habra mucho tiempo para hablar de esto.

### Functional programming:  Paradigma de programacion  funcional.

Como el nombre lo indica se basa en funciones. Este paradigma tiene unas caracteristicas que vamos a definir a continuacion.

* Las funciones son **puras**. Es decir, dados ciertos argumentos, la funcion debe retornar el **mismo** valor para los **mismos argumentos**. 

* No tiene efectos colaterales (no side effects). No modifican las variables. Las variables son inmutables.

* **Recursion**: no existen cilclos "while", "for" todo se simula con recursion.
