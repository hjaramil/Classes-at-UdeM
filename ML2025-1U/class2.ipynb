Definamos la inversa de una matriz.

**Definicion 2.1.15** (inversa de una matriz): Sea $A$ una matriz cuadrada $n \times n$ (compleja o real).  Decimos que $B$ es la **inversa** de $A$ si
$AB =  BA = I$, escribimos $B=A^{-1}$.  Decimos que $A$ es **no singular**. Si no existe la inversa, se dice que $A$ **singular**.


Para resolver la pregunta de si $A^* A = A A^*$ vamos a ver un par de teoremitas (del algebra lineal)

**Teorema 2.1.7** (determinantes y inversas)
1. Si $A,B$ son matrices cuadradas, entonces $\det(AB)=\det(A) \det(B)$.
2. Si $\det(A) \ne 0$ entonces la inversa de $A$ existe. Es no singular.

El teorema que probamos enseguida muestra cuando dos matrices (cuyo producto sea la identidad)  conmutan.

**Teorema 2.1.8**: Si $AB=I$, y $A,B$ son cuadradas entonces $AB=BA=I$, y $B$ es la inversa de $A$.

Prueba: El $\det(I)=1$. Como $\det(AB)=\det(A) \det(B)=1$, entonces $B$ tiene inversa por que $\det(B) \ne 0$. Existe $B^{-1}$.
Como $AB=I$ entonces

$$B= B (AB) = (BA) B.  \tag{3} $$

Ahora multiplicamos (3) a ambos lados por $B^{-1}$

$$BB^{-1}=  (BA) B B^{-1}.  $$
$$I =  (BA) I.  $$
$$I =  (BA)   $$
De forma que

$$AB=BA = I.  $$

De forma que, de verdad, si $A$ es ortogonal $A^* A = A A^* = I$.


**Teorema 2.1.9**: Si $A,B$ son no singulares entonces la inversa del producto es el producto de las inversas pero al reves.

$$ (AB)^{-1} = B^{-1} A^{-1}. $$

Prueba:

$$(B^{-1} A^{-1})(AB) =  B^{-1} (A^{-1} A) B = B^{-1} B = I. $$

Definamos espacios nulo y rango

**Definicion 2.1.17**: (espacio rango y espacio nulo  o kernel):
Dada una matriz $A$ $m \times n$. Definimos el **rango** de $A$ como el espacio generado por las columnas de $A$. Es decir si $v_i$ son las columnas de $A$ el rango es el espacio de todas las combinaciones lineales $\sum_{i=1}^n \alpha_i v_i$. Con $\alpha_i \in \mathbb{C}$. Al rango de $A$ que tambien se le llama
**espacio columna**. El **espacio nulo** o **kernel** es el conjunto de todos los vectores $x$ tales que $Ax = 0$.

## Autovalores/Autovectores
Tambien se conocen como valores propios/vectores propios (en ingles Eigenvalues/Eigenvectors, del Aleman eigne=propio)

Si existe un vector $x$ tal que para la matriz cuadrada $A$

$$A x = \lambda x \tag{4} $$
decimos que $\lambda$ es un autovalor de $A$ y $x$ el autovector correspondiente.

Queremos que $x \ne 0$. Este es trivial
Como hallamos los autovalores/vectores?
de la ecuacion (4)

$$A x - \lambda x  = 0  $$
$$A x - \lambda I x  = 0  $$
$$(A - \lambda I) x  = 0  \tag{5} $$
Esto lo que  quiere decir (algebra lineal) es que el determinante de la matriz $A - \lambda I$ es 0.

El algoritmo es: Resolver la ecuacion

$$p(\lambda)  = \det(A - \lambda I) = 0 $$
para $\lambda$. Siempre se garantiza la existencia de autovalores por que todo polinomio tiene por lo menos una raiz (teorema fundamental del algebra).
El polinomio $p(\lambda)$ se llama **polinomio caracteristico**. Una vez hallado un $\lambda$ se usa la ecuacion (5) para hallar los autovectores.


Para que quede mas claro el concepto voy a hacer unos ejemplos.

**Ejemplo 2.2.1**: Encuentre los autovalores/vectores de la siguiente matriz

$$ A = \begin{pmatrix} -5 & -3 \\ - 6 & 2 \end{pmatrix}. $$

Hallemos el polinomio characteristico $p(\lambda)$.


\begin{eqnarray}
p(\lambda ) =  \det( A - \lambda I) &=& \det
\begin{pmatrix}
-5 - \lambda  & - 3 \\
-6 & 2 - \lambda
\end{pmatrix} \\
&=& (-5 - \lambda)(2 - \lambda) - 18 \\
&=& -10 + 3 \lambda + \lambda^2 - 18 \\
&=& \lambda^2 + 3 \lambda - 28 \\
&=& (\lambda + 7)(\lambda - 4) = 0.
\end{eqnarray}
Entonces las raices son $\lambda = -7$ y $\lambda = 4$.
Esta es la primera parte del algoritmo. Para la segunda parte tomamos cada $\lambda$ y hallamos los auto-vectores correspondientes.

1. $\lambda=-7$.  Debemos resolver el sistema $(A - \lambda I)x=0$.
De forma que vemos que hay infinitas soluciones. Por esta razon a veces se buscan $x$ que sean normalizados $\| x \|=1$. La ecuacion es "degenerada" las filas no son linealmente independientes.

Podemos escoger la que nos convenga.  Veamos

\begin{eqnarray}
\left \{
    \begin{pmatrix} -5 & - 3 \\ -6 & 2 \end{pmatrix}
    - \begin{pmatrix} -7  & 0  \\ 0 & -7 \end{pmatrix}
    \right \}
    \binom{x}{y} &=& \binom{0}{0} \\
    \begin{pmatrix} 2 & -3 \\ -6 & 9\end{pmatrix}
    \binom{x}{y} &=& \binom{0}{0} \\
\end{eqnarray}

Tomemos la primera fila
$$2 x - 3y = 0 $$
Hay infinidad de soluciones. Escojamos $x=3,y=2$ de forma que

$$u = \binom{x}{y}  = \binom{3}{2} $$
Es un autovector correspondiente al autovalor $\lambda = -7$.






2. $\lambda=4$
El sistema queda

$$ \begin{pmatrix} -9 & -3 \\ -6 &-2 \end{pmatrix}  \binom{x}{y} = 0 $$

Tomemos la primera

$$-9x - 3 y = 0 $$
$$3x  + y = 0 $$
Soluciones (infinitas), escojamos $x=-1$, $y=3$.
De form que un autovector correspondiente a $\lambda=4$ es

$$3x + y = 0 $$
Despejemos $y$, $y=-3x$.
La solucion es una linea (infinitos puntos)

$$ \binom{x}{-3x} $$
Arbitrariamente yo puede decir $x=1$, entoncese
$$ \binom{1}{-3} $$


```python
import numpy as np
A = np.array( [[-5, -3],[-6, 2]])

# find the eigenvalues of A
eigenvalues, eigenvectors = np.linalg.eig(A)

print(eigenvalues)
print(eigenvectors)
```
