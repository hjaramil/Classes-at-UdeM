**Ejemplo 2.2.2**: Encuentre los autovalores/vectores de

$$A = \begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix}  $$

Queremos explorar a que conjunto mapea la matriz sobre el cuadrado de vertices $$(1,1), (-1,1), (-1,1)), (1,-1) $$ y el circulo unitario $x^2 + y^2 = 1$ $(\cos \theta, \sin theta)$ (parametrizado aca por $\theta$.

1. Autovalores
\begin{eqnarray}
p(\lambda) =\det(A - \lambda I) = \det
\begin{pmatrix}
1 - \lambda & 1/2 \\ 1/2 & 1 -\lambda
\end{pmatrix}  \\
(1-\lambda)^2 - \frac12^2 \\
\left (  1 - \lambda - \frac12  \right )
\left (  1 + \lambda  + \frac12  \right )  \\
(1/2 - \lambda)(3/2 - \lambda).
\end{eqnarray}
Dos valores propios.  $\lambda=1/2, \lambda = 3/2$.

2. Autovectores:
    * Para $\lambda_2=1/2$.

    $$\frac12 x + \frac12 y = 0   $$
    La solucion es $y=-x$. Los autovectores son

    $$v = \binom{x}{-x} $$
    en particular para $x=1$

    $$v= \binom{1}{-1} $$

    * para $\lambda_1=3/2$:
    $$-\frac12 x + \frac12 y = 0   $$
    La solucion es $y=x$. Los autovectores son

    $$v = \binom{x}{x} $$
    en particular para $x=1$
    $$v= \binom{1}{1} $$

Ahora veamos el mapeo $A x$ del cuadrado indicado

$$\begin{pmatrix}1 & 1/2 \\ 1/2 & 1 \end{pmatrix} \binom{1}{1} = \binom{3/2}{3/2}  $$

$$\begin{pmatrix}1 & 1/2 \\ 1/2 & 1 \end{pmatrix} \binom{-1}{1} = \binom{-1/2}{1/2}  $$

$$\begin{pmatrix}1 & 1/2 \\ 1/2 & 1 \end{pmatrix} \binom{-1}{-1} = \binom{-1/2}{1/2}  $$

$$\begin{pmatrix}1 & 1/2 \\ 1/2 & 1 \end{pmatrix} \binom{1}{-1} = \binom{1/2}{-1/2}  $$

Grafiquemos estos cuatro puntos y analizamos lo que pasa.

El tamanho de los vectores de partida es

$$ \| u \| = \| v \| = \sqrt{1^2 + 1^2} = \sqrt{2} $$

El tamanho de los vectores de llegada ($Ax$) es

$$\| u ' \| = \sqrt{ (3/2)^2 + (3/2)^2} = \sqrt{9/2}=\frac{3}{\sqrt{2}} $$
$$\| v ' \| = \sqrt{ (1/2)^2 + (1/2)^2} = \sqrt{1/2}=\frac{1}{\sqrt{2}} $$

Ahora vamos a ver analitcamente y luego verificar numericamente que los autovalres son factores de escala del operador $A$.

$$\frac{\| u' \|}{ \| u \|} =\frac{\| A u \| }{\| u \| } = \frac{\| \lambda_1 u \| }{\| u \| } = \frac{ |\lambda_1 | \| u \|}{\| u \|} = | \lambda_1 | $$

$$\frac{\| v' \|}{ \| v \|} =\frac{\| A v \| }{\| v \| } = \frac{\| \lambda_2 v \| }{\| v \| } = \frac{ |\lambda_2 | \| v \|}{\| v \|} = | \lambda_2 | $$

Se puede verificar numericamente que

$$\lambda_1 = \frac{\| u' \| }{\| u \|} = \frac32 $$
$$\lambda_2 = \frac{\| v' \| }{\| v \|} = \frac12 $$

Que pasa cuando mapeamos el ciculo unitario por media de $A$

$$ \begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix} \binom{\cos \theta}{\sin \theta} = \binom{\cos \theta + \frac12 \sin \theta}{\frac12 \cos \theta + \sin \theta} $$
con $\theta \in [0, 2 \pi)$.

**Teorema 2.2.2**: Sea $A$ una matriz autoadjunta, entonces

* Los autovalres son reales (no complejos)
* Los autovectores correspondientes a autovalores distintos son ortogonales
(ver por ejemplo el Ejemplo 2.2.2)

Prueba.
Mostremos primero que los autovalres son reales.
Preliminar (Lema)
Muestre primero que

$$\langle A x, y \rangle = \langle x, A^* y \rangle \tag{2} $$
Prueba:

$$\langle Ax,  y \rangle = (Ax)^* y = x^* A^* y = x^* (A^* y) = \langle x, A^*y \rangle  $$

Usemos esto para verificar que los autovalores de la matriz $A$ son **reales**.

Asuma $\lambda_j$ un autovector de $A$, podemos sin perdida de generalidad asumir que el autovector correspondiente $u_j$ es unitario. Es decir $\| u_j \|=1$. Calculemos la forma cuadratica $\langle A u_j, u_j \rangle $.

Por un lado.
$$\langle A u_j, u_j \rangle = \langle \lambda_j u_j, u_j \rangle = \overline{\lambda_j} \langle u_j, u_j \rangle = \overline{\lambda_j} \| u_j \|^2 = \overline{\lambda_j}.  \tag{3} $$


Por otro lado por (2)  $\langle A u_j, u_j \rangle = \langle u_j, A^* u_j \rangle$

Entonces, de la ecuacion (3) y como a es autoadjunta

$$\overline{\lambda_j} = \langle A u_j, u_j \rangle = \langle u_j, A^* u_j \rangle =  \langle u_j, A u_j \rangle = \langle u_j, \lambda u_j \rangle = \lambda \langle u_j, u_j \rangle = \lambda $$

Es decir $\overline{\lambda} = \lambda$, o sea que $\lambda \in \mathbb{R}$.

Probemos que si $a+\mathrm{i} b = a - \mathrm{i} b$, entonces $b=0$

$$a + \mathrm{i}b = a - \mathrm{i} b $$
$$ \mathrm{i}b =  - \mathrm{i} b $$
$$ \mathrm{i} (b + b) = 0 $$
$$ 2 b \mathrm{i} =0$$
$$ b=0$$.

Listo, los autovalores son reales.
Probemos ahora que los autovectores son ortogonales.
Sean $u,v$ dos autovectores correspondientes a autovalores distintos $\lambda_1, \lambda_2$

Entonces
$$\langle A u, v \rangle = \langle \lambda u , v \rangle = \lambda_1 \langle u , v \rangle   \tag{4} $$
Ahora, como $A=A^*$

$$\langle A u, v \rangle = \langle u, Av \rangle = \langle u, \lambda_2 v \rangle = \lambda_2 \langle u, v \rangle \tag{5}  $$

Comparando (4) con (5) tenemos que

$$\lambda_1 \langle u , v \rangle = \lambda_2 \langle u , v \rangle $$
Es decir

$$(\lambda_1 - \lambda_2) \langle u, v \rangle = 0 $$
Como los autvalores son distintos $\lambda_1 - \lambda_2 \ne 0$.

O sea que $\langle u, v \rangle = 0$.
Y $u,v$ son ortogonales.

Esto nos ayuda a encontrar una factorizacion (o descomposicion) interesante de la matriz $A$.

Sea $A \in \mathbb{C}^n$, Asumamos por ahora que **todos** los autovalores son distintos, entonces los autovectores $x_i$ (normalizados) son ortogonales entre si y por tal linealmente independientes.  Tenemos

$$ A x_i = \lambda_i x_i \quad , \quad i=1,2, \cdots, n $$
Como son ortogonales entre si

$$ x_i^* x_j = \delta_{ij}  $$

Entonces
$$x_i^* A x_j = x_i^* \lambda_j x_j = \lambda_j \delta_{ij} $$

El triple producto

$$Q^* A Q = \Lambda $$

Donde $Q$ es una matriz cuyas columnas son $x_i$, $\Lambda$ es una diagonal con autovalores $\lambda_i$.

Si multiplicamos esto a izquierda por $Q$ y a derecha por $Q^*$



$$ Q Q^* A Q Q^* = Q \Lambda  Q^*$$
ahora como $Q Q^* = Q^* Q = I$
o sea que
$$  A  = Q \Lambda  Q^*$$

Esta es la descomposion de $A$ en el producto de dos matrices ortogonales y una matriz diagonal.

Este teorema asume que los autovalores sean todos distintos, pero si aun hay repetidos el teorema sigue siendo valido. La prueba esta en las notas que les comparti. La dejo de tarea de lectura. El teorema es:

**Teorema 2.2.4**: (teorema de descomposicion espectral de operadores (matrices) autoadjuntos).
Sea $A$ una matriz autoadjunta. Entonces $A$ se puede escribir como el producto de tres matrices, dos ortogonales y una diagonal como

$$  A  = Q \Lambda  Q^*$$
donde $Q$ es una matrix cuyas columnas son los autovectores de $A$ y $\Lambda$ es una matriz diagonal donde $\lambda_{ij}=\delta_{ij} \lambda_j$ son los autovalores de $A$.

Una aplicacion inmediata de este teorema es la simplificacion de formas cuadraticas.

Por ejemplo, sobre los reales, para $A \in \mathbb{R}^{n \times n}$ autoadjunta (simetrica)




$$x^T A x = \sum_{i=1}^n \sum_{j=1}^n x_i a_{ij} x_j
= \sum_{i=1}^n \sum_{j=1}^n  a_{ij} x_i x_j$$
Cuantas operaciones hay aca?  multiplicaciones del orden $n^2$, sumas $n^2$
Si $n$ es grande, por ejemplo $n=1000$, $n^2 = 10^6$.

Si $A$ es autoadjunta $A= Q \Lambda Q^*$
Llamemos $y=Q^T x$, $y^T = x^T Q$,


$$ x^T A x = x^T Q \Lambda Q^T x = y^T \Lambda y $$
Esta ultima operacion es

$$ x^T A x = \sum_{i=1}^n \lambda_i y_i^2 $$



Hemos visto muchas formas de multiplicar matrices. El proximo teorema extiende estas formas.

**Teorema 2.2.5**: Si $A$ es una matriz $m \times n$. $B$ es una matriz $n \times p$, y $x$ es un vector de $n$ componentes, y $\Lambda$ es una diagonal con $\lambda_i$ en la diagonal entonces

1.

$$AB = \left [  AB_1 | AB_2 | \cdots | AB_p  \right ]  \tag{9} $$
donde $B_j$ es la columna $j$ de $B$. Lo que esto dice es que para hallar la columna $j$ de $AB$ no necesitamos hacer todo la multiplicacion $AB$ y extraer la columna sino que solo basta con multiplicar $A$ por $B_j$.

2.
$$A x = \sum_{i=1}^n x_j A_j  \tag{10}$$
donde $A_j$ es la columna $j$ de $A$.
Claramente si variamos $x$ en todo el espacio $\mathbb{R}^n$ o $\mathbb{C}^n$ obtenemos el rango (espacio columna) de la matriz $A$. Es decir $\mathcal{R}(A)$.

3. Si $p \ge n$

$$ A \Lambda  = [ \lambda_1 A_1 | \lambda_2 A_2 | \cdots | \lambda_n A_n | 0_{n=1} | \cdots 0_{p} ] \tag{11} $$
donde hay $p-n$ columnas de ceros. Si $n>p$

$$ A \Lambda  = [ \lambda_1 A_1 | \lambda_2 A_2 | \cdots | \lambda_p A_p ] \tag{12}$$

4. Si $u_i$ es vector columna de la matriz $U$ que es $m \times m$, y $b \in \mathbb{C}^{m}$, entonces

$$U^* b = \begin{pmatrix} u_1^* b \\ u_2^* b \\ \vdots \\ u_m^* b \end{pmatrix} \tag{13} $$

**Prueba**:

1. Mostremos
Verificamos (9).
Llamemos $C$ al producto de $A$ por $B$, $C=AB$. Entonces

$$c_{ij} = \sum_{k=1}^n a_{ik} b_{kj}. \tag{14} $$

Ahora consideremos la columna $j$ del lado derecho (10).

$$(A B_j) = \sum_{k=1}^n a_{ik} b_{kj} $$
que es lo mismo de  (14) lo que prueba (10).


2. Probamos las (10).
Note que la componente $i$ de ambos lados es

$$(Ax)_i = \sum_{j=1}^n a_{ij} x_j = \left (  \sum_{j=1}^n x_j A_j  \right )_i $$
donde $A_j$ es la columna $j$ de $A$. Si se expanden ambos lados se ve la igualdad.

3. De la (9)
Verificamos (11)
Asumamos que $p<n$
\begin{eqnarray}
A \Lambda &=& [ A \Lambda_1 | A \Lambda_2 | \cdots | A \Lambda_p]  \\
\end{eqnarray}
Ahora, usando (10) con

$$x=\Lambda_1 = \begin{pmatrix} \lambda_1 \\ 0 \\ 0 \\ \cdots \\ 0 \end{pmatrix}  $$
encontramos

$$ A \Lambda_1 = \lambda_1 A_1 $$

$$ A \Lambda_i = \lambda_i A_i $$
donde $A_i$ es la $i$ columna de $A$. asi mismo
con $i=1,2 \cdots, p$ de forma que obtenemos el resultado (12).

4. Finalmente probamos (13). Tomemos como $i$ la componente (fila) $i$ del
vector $U^* b$, es decir $(U^* b)_i$ entonces

$$ (U^* b)_i = \sum_{j=1}^m \overline{u_{ji}} b_j = \langle u_i , b \rangle = u_i^* b $$

Es decir

$$U^* b = \begin{pmatrix} u_1^* b \\ u_2^* b \\ \vdots \\ u_m^* b \end{pmatrix} $$



Ya vimos una forma de **diagonalizar** mediante el uso del teorema espectral. Veamos otra forma donde no se necesita que la matriz sea autoadjunto.
Asumamos que la matriz $A$ en $\mathbb{C}^{n \times n}$ es tal que sus **autovectores son linealmente independientes** (con autovalores $\lambda_i$). Entonces veamos
Si los autovectores de esta matriz son $P_i, i=1,2, \cdots, n$ entonces

$$A P_i = \lambda_i P_i  \quad, \quad i=1,2, \cdots, n$$
Podemos usar el 2.2.5. numeral tres y escribir

$$A P = P \Lambda  \tag{15} $$
donde $\Lambda$ es una matriz diagonal con los autovalores de $A$ en la diagonal.

Podemos despejar $A$ de la ecuacion (15), multiplicando a derecha por $P^{-1}$

$$A P P^{-1} = P \Lambda P^{-1}$$
pues $P$ tiene inversa por que sus columnas son linealmente independientes (es de rango $n$).


$$A = P \Lambda P^{-1} \tag{16}$$

Esto nos dice
**Teorema 2.2.6**:  Que una matriz $A$ con autovectores linealmente independientes se puede descomponer en el producto de tres matrices
$$A = P \Lambda P^{-1}$$
donde $P$ es la matriz con columnas autovectores de $A$ y $\Lambda$ es la diagonal con autovalroes de $A$.
Decimos entonces que $A$ es **diagonalizable** (definicion)

Aplicacion de esto a la exponenciacion de matrices.

**Teormea 2.2.7**:  Si $A$ es diagonalizable

$$A^k = P \Lambda^k P^{-1}.  \tag{17} $$

La prueba de esto es por induccion
(lo explique en palabras en la clase)

* Caso $k=1$
$$A = P \Lambda P^{-1}. $$
que se cumple por (16).

* Asumamos que es valido para $k$ y probamos que esto implica que tambien es valido para $k+1$. Es decir asumamos (17)
$$A^k = P \Lambda^k P^{-1}$$.  

Debemos probar
$$A^{k+1} = P \Lambda^{k+1} P^{-1}.  \tag{17}$$

Veamos
\begin{eqnarray}
A^{k+1} &=& A^k A \\
&=& (P \Lambda^k P^{-1}) A \\
&=& (P \Lambda^k P^{-1}) (P \Lambda P^{-1}) \\
&=& P \Lambda^k ( P^{-1}P) \Lambda P^{-1} \\
&=& P \Lambda^k ( I) \Lambda P^{-1} \\
&=& P \Lambda^k \Lambda P^{-1} \\
&=& P \Lambda^{k+1} P^{-1} \\
\end{eqnarray}
probamos (17) y por tanto el teorema es valido para los enteros $k=1,2, \cdots, \infty$ por el

[principio de induccion matematica](https://es.wikipedia.org/wiki/Inducci%C3%B3n_matem%C3%A1tica).

**Definicion 2.2.3**(Positiva definda, negativa definida): Una matriz autoadjunta $A$ es **positiva-definida** si todos los autovalores son positivos. Negativa definda si todos los autovalores son negativos.
En la literatura se definie positiva definida si, para todo $x \in \mathbb{C}^n$

$$x^* A x > 0  $$
En la literatura se definie negativa definida si, para todo $x \in \mathbb{C}^n$
$$x^* A x < 0  $$
Se puede probar que las dos definiciones son equivalentes  Es decir que si se asumen las ultimas propiedades se obtienen las primeras y viceversa. Creo que esto lo tengo en la proxima tarea.
