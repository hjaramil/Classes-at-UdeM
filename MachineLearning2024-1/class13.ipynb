# Feature Normalizacion (normalizacion de parametros)
Si las columnas de la matriz de datos $X$ tienen valores muy
diferentes entre columnas, se deben normalizar. Esto tendria las siguientes ventajas.

1. Incrementa la velocidad de convergencia (proceso) con menos "zigzags". Por que las curvas, o superficies, o hipersuperficies, son mas "esfericas"
2. Incrementa la estabilidad. El "number condition" ($\kappa$)
se acerca mas a uno, por que es el radio entre el autovalor mas grande y el mas pequenho. Si las columnas estan bien balanceadas, los autovlares se acercan mas entre si.

2. Elimina el sesgo. Columnas con valores muy grandes pueden "empujar" el resultado hacia ese lado.


Que es normalizacion? Es dividir las columnas por algunos escalares que las balanceen lo mejor posible. Usted puede decir: pero esto cambia el problema, y ya estamos resolviendo otro problema.


La respuesta es "si, pero no"
Explico. El sistema original


$$ X \theta = y .$$
Al normalizar las columnas, tenemos un sistema nuevo


$$X_n \theta_n = y \tag{1} $$
el $y$ no se cambia. Este nuevo problema (1) es mas estable.
Cual es truco,

* se guardan (save) los valores con los cuales se normalizaron las columnas.

* Se resuelve el problema $X_n \theta_n = y$. Se halla $\theta_n$

* Se entregan datos nuevos, estos datos se normalizan con los **mismos** valores encontrados en el numeral anterior. Para obtener una mueva matriz $X_n$.
La nueva hipotesis es $h(x, \theta) = X_n \theta_n$.
Esto me genera precios compatibles con los $y$ del problema original.

## Metodos de normalizacion:

* Estandarizacion (Standardization): Se normalizan los datos de forma que las columnas de $X$ tienen media 0 y desviacion estandard 1. Tome cualquier columna $X_j$ (excepto la primera que es de unos) y aplique la siguiente ecuacion

La columna normalizada es


$$X_j \leftarrow  \frac{X_j - \mu_j}{\sigma_j} .$$

donde

$$\mu_j = \frac{\sum_{i=1}^m x_{ij} }{m} .$$

y
$$\sigma_j = \sqrt{ \frac{\sum_{i=1}^m (x_{ij}- \mu_j)^2}{m}} .$$

Luego de "standardization" cada columna (excepto la primera) tiene media 0 y desviacion estandard 1.


* Normalizacion por la media: Esta se obtiene removiendo la media de cada columna y normalizando (dividiendo) por el "rango" de la columna (diferencia entre el maximo y el minimo)
mas concretamente


$$X_j \leftarrow \frac{X_j - \mu_j}{\max_i \{ x_{ij} \} - \min_i \{ x_{ij} \} }  \quad i=1,2, \cdots, m.$$

* Scalada por Min-Max

$$X_j \leftarrow \frac{X_j - \min_i \{ x_{ij} \}}{\max_i \{ x_{ij} \} - \min_i \{ x_{ij} \} }  \quad i=1,2, \cdots, m.$$




* Por ultimo, normalizacion por la norma del vector (vector norm normalizacion)

$$X_j \leftarrow \frac{X_j}{ \| X_j \|} .$$

# Regresion polinomial (que tambien es lineal)

Veamos: Considere un polinomio en una variable


$$h(\theta ,x) = \theta_0 + \theta_1 x + \theta_2 x^2 + \cdots + \theta_n x^n .$$
Esta ecuacion la puede escribir de la forma


$$h(\theta, x) = X \theta =\langle X, \theta \rangle = X \cdot \theta = .$$
donde $X=(1, x, x^2 , \cdots , x^n)$, $\theta=(\theta_0, \theta_1, \cdots, \theta_n)^T.$

Si tenemos $m$ muestras podemos repetir esta ecaucion muchas veces, para muestra de la siguiente forma


\begin{eqnarray}
\begin{pmatrix}
1 & x_1 & x_1^2 & \cdots & \cdots & \cdots & x_1^n \\
1 & x_2 & x_2^2 & \cdots & \cdots & \cdots & x_2^n \\
\vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots \\
\vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots \\
1 & x_m & x_m^2 & \cdots & \cdots & \cdots & x_m^n \\
\end{pmatrix}
\begin{pmatrix}
\theta_0 \\
\theta_1 \\
\vdots \\
\vdots \\
\theta_n
\end{pmatrix}
=
\begin{pmatrix}
y_1 \\
y_2 \\
\vdots \\
\vdots  \\
y_m
\end{pmatrix}
\end{eqnarray}




Si $m=n+1$ esta matriz es cuadrada y se llama la matriz de

[la matriz de Vandermonde](https://en.wikipedia.org/wiki/Vandermonde_matrix)

Esta matriz es justo la que se usa para interpolacion de **Lagrange**.
En este punto **muy especifico**

$$ interpolacion = regresion .$$

El 99.9999999999999999% de las veces esto no ocurre

En general no queremos ajustar muy precisamente los datos a un polinomio por que generamos **sobreajuste** (mas adelante explico este concepto.

En el caso de la matriz cuadrada (la Vandermonde $V$) tenemos

$$ \theta = V^{-1} y .$$


### Regresion polinomica en varias variables.

Escribamos un ejemplo. Polinomio en dos variables (numero de cuartos y area de la casa), de orden cuadratico (2)


$$ h(\theta_0, \theta_1, \theta_2, \theta_3, \theta_4, \theta_5, x_1, x_2) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \theta_3 x_1 x_2  + \theta_4 x_1^2 + \theta_5 x_2^2 .$$




La matriz para este polinomio



\begin{eqnarray}
X =
\begin{pmatrix}
1 & x_{11} & x_{12} & x_{11} x_{12} & x_{11}^2 & x_{12}^2 \\
1 & x_{21} & x_{22} & x_{21} x_{22} & x_{21}^2 & x_{22}^2 \\
\vdots & \vdots & \vdots & \vdots & \vdots &  \vdots \\
\vdots & \vdots & \vdots & \vdots & \vdots &  \vdots \\
\vdots & \vdots & \vdots & \vdots & \vdots &  \vdots \\
1 & x_{m1} & x_{m2} & x_{m1} x_{m2} & x_{m1}^2 & x_{m2}^2 \\
\end{pmatrix}
\end{eqnarray}

### Regresion polinomial en varias variables en general
Un caso cubico, donde evito los cuadrados $x_1^2$, $x_2$.

$$h(\theta) = p_{\theta}(x_1, x_2) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \theta_3 x_1 x_2 + \theta_4 x_1^2 x_2 + \theta_5 x_1 x_2^2 + \theta_5 x_1^3 + \theta_7 x_2^3 .$$

La matriz para este polinomio:

\begin{eqnarray}
X =
\begin{pmatrix}
1 & x_{11} & x_{12} & x_{11} x_{12} & x_{11}^2 x_{12} & x_{11} x_{12}^2  x_{11}^3 & x_{12}^3 \\
1 & x_{21} & x_{22} & x_{21} x_{22} & x_{21}^2 x_{22} & x_{21} x_{22}^2  x_{21}^3 & x_{22}^3 \\
\vdots & \vdots & \vdots & \vdots & \vdots &  \vdots &\vdots \\
\vdots & \vdots & \vdots & \vdots & \vdots &  \vdots &\vdots \\
\vdots & \vdots & \vdots & \vdots & \vdots &  \vdots &\vdots \\
1 & x_{m1} & x_{m2} & x_{m1} x_{m2} & x_{m1}^2 x_{m2} & x_{m1} x_{m2}^2  x_{m1}^3 & x_{m2}^3 \\
\end{pmatrix}
\end{eqnarray}

Ojo, no solamente podemos hablar de interpolacion polinomica, podemos pensar de forma general con una base $(1, x, \sqrt{x})$
donde $$h(\theta,x) = \theta_0 + \theta_1 x + \theta_2 \sqrt{x} .$$

Igual, en vez de $\sqrt{x}$ podemos pensar en logaritmo, exponencial, trigonometrica, etc.




## Overfitting/Underfitting (sobreajuste/subajuste)
image.png


Se usa la funcion $\log x \sin x$ para generar los puntos azules. Hacemos regresion (polinomial) con polinomios de grado 3, 4, 5. Ninguno de ellos se ajusta tan bien. Que tal polinomios de mayor grado.



image.png

* En la practica tenemos muchos datos $m \gg n$, el problema es mas de regresion que de interpolacion.

* El sobreajuste ocurre cuando se tienen muchos "features".
En el caso de la regresion polinimial son (pseudo-features) y
la potencia es grande. Los polinomios de grado alto oscilan muchos (tienen muchas variabilidad, varianza). Mas de los polinomios. Al final de los polinomios aparece un efecto numerico llamado "Runge Phenomenon". Se disparan hacia arriba o hacia abajo.

* Evitar el "overfitting" hace que la curva sea mas suave.

Que problemas produce el "overfitting"

* Nos pegamos mucho al ruido. Aprendemos del ruido.
* Debemos prevenir oscilaciones muy grandes y de alta frecuencia.
* Computacionalmente es mas costoso implementar un algoritmo con muchos "features"

* Finalmente, crear modelos que son inaceptables no es lo que buscamos en ML.


Como evitar el overfitting?

* reducir el numero de atributos (features)
* regularizacion (esto lo vemos mas adelante)
* Usar distintos modelos de ML con combinaciones de datos (aumentar los datos)
* Dividir los datos en datos de entremiento y datos de prueba.
Este ultimo paso, lo vamos a discutir en detalle, es mas un diagonstico de "overfitting".


Lo contrario a **overfitting** es **underfitting**.

* demasiado suaves las "curvas" (lineas rectas o muy suaves)
* muy poca varianza, no oscilaciones
