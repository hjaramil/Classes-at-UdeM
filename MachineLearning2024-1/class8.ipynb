# Repaso de la ultima clase
Vimos calculo matricial. En este proceso definimos el gradiente de una matriz o de una campo escalar.  Entre estos definimos

$$f(x) = \| Ax - b \|^2 . \tag{1} $$
Por que es esto importante.

$$Ax = b .$$
Este sistema es central en el algebra lineal.
Si $x$ resuelve la ecuacion $Ax - b = 0$
pero si no $Ax - b = E$, $E$ error, y construimos la funcion **objetivo**, (1)
la llaman funcion de costo, de perdida (loss function).

La minimizacion de estas funciones (en el continuo) se da a traves del gradiente.

Vimos que

$$ \nabla f = A^T A x - A^T b .$$
Es decir que si $\nabla f = 0$.
Esto es cierto si

$$ A^* A x = A^* b .$$
A esto se le llaman las ecuaciones normales.
Este problema esta en el centro de la **regresion lineal**.

Si $A^*A$ tiene inversa entonces

$$x = (A^* A)^{-1} A^*b .$$


# RegularizaciÃ³n. (Regularized least squares)
El problema que nos interesa en el momento es

$$ Ax = b .$$
donde $A \in \mathbb{R}^{m \times n}$, $b \in \mathbb{R}^m$ y
$x \in \mathbb{R}^n$.

En muy contados casos este problema tiene solucion. Si la matriz $A$
tiene inversa (no singular) entonces si tiene solucion

$$ x = A^{-1} b.$$
Sin embargo en **gran** mayoria de los casos $A$ ni siquiera es cuadrada.

Ej: $x + y = 5$.
\begin{eqnarray}
(1   \; 1) \binom{x}{y} = 5
\end{eqnarray}
En este ejemplo $A=(1 \; 1)$. $A$ no tiene inversa. Aun siendo cuadrada
la mayoria no tiene inversa.

$$ A = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix} .$$
$A$ no tiene inversa.

Si $A$ no es cuadrada, puden resultar varias posibilidades.

1. Que el sistema $Ax=b$ sea **sobre-derminado**, es decir que hay mas ecuaciones independientes que incogitas.

Ej:
\begin{eqnarray}
x=5 \\
x=2
\end{eqnarray}
Matricial
\begin{eqnarray}
\begin{pmatrix} 1 \\ 1 \end{pmatrix} x = \begin{pmatrix} 5 \\ 2 \end{pmatrix}.
\end{eqnarray}
En este caso $A=\binom{1}{1}.$

En este caso el espacio nulo $\mathcal{A} = \{ 0 \}$ pero el rango
no cubre todo el espacio .

2. El sistema esta **subdeterminado**. En este caso hay mas variables que ecuaciones (linealmente independientes)


$$ x + y = 5$$.
En este caso el espacio nulo noes el trivial $\{ 0 \}$.
Recuerden el espacio nulo de $A$ es el conjunto de los $x$ tales
que $Ax=0$.  En este caso $A=(1  \; 1)$.
El espacio nulo es $\binom{x}{-x}$


$$ (1 \; 1) \binom{x}{-x} = x -x = 0 .$$

Por que la solucion no es unica. Asuma una solucion $x_0$ de
$Ax_0 = b$, sea $x_n \in \mathcal{N}$, entonces


$$ A(x_0 + x_n) = Ax_0 + A x_n = A x_0 + 0 = b.$$


Queremos hallar una solucion unica y **estable** de $Ax=b$.
Una serie de matematicos, entre ellos Tikhonov (Ruso) plantearon el siguiente problema.  Resolver la fucion objetivo.



$$ J(x) = \| Ax - b \|^2 + \| \Gamma x \|^2 \tag{1a}.$$
donde $\Gamma$ es una matriz **positiva definida**.

Si usamos calculo matricial, como en la clase anterior (hallando el gradiente de $f$, $\nabla f$) e igualandolo a cero (ejercicio para el estudiante), encontramos,

$$ (A^* A + \Gamma^* \Gamma) x = A^*b .$$
Ahora, los autovalores de $A^*A $ son positivos o 0. Pero como
$\Gamma$ es positiva definida, los autovalores de $\Gamma^* \Gamma$
son positivos y por eso los de la suma ($A^*A + \Gamma^* \Gamma)$
tambien son positivos y la matriz es invertible.


resolviendo la ecuacion resultante

$$x = (A^*A + \Gamma^* \Gamma)^{-1} A^* b . \tag{2}$$

Es comun ver que $\Gamma = \sqrt{\lambda} I$, y en este caso la solucion es


$$x = (A^*A + \lambda I)^{-1} A^* b .$$ donde $\lambda>0$.
A este problema en estadistica lo llaman **ridge regression**.

Veamos como obtener la ecuacion (2) de forma heuristica (sin usar
calculo, hallar el gradiente e igualarlo a cero)

Comenzamos con $Ax=b$.
Multiplicamos a ambos lados por $A^*$


$$ A^* Ax = A^* b . \tag{3}$$
Si $A^*A$ tiene inversa

$$x = (A^* A)^{-1} A^* b .$$
Sin embargo si hay autovalores de $A^*A$ que son 0, entonces toca
"empujar el espectro al lado positivo de la recta real" en la ecuacion (3)

$$ (A^* A + \lambda I) x = A^* b . \tag{4}$$
donde $\lambda >0$.  En la practiva $\lambda$ es pequenho,
si $\lambda \to \infty$, entonces $x \to 0$.
donde
$$ x = (A^* A + \lambda I)^{-1} A^* b . \tag{5}$$
Como $\lambda >0$ la solucion (5) tiene dos caracteristicas.

1. Siempre existe
2. Es unica.

Ojoooo: Es posible que todos los autovalores de $A^* A$ sean positivos,
pero los mas peuquenhos son del orden $10^{-8}$. Esto genera **inestabilidad**.

En la funcion objetivo (1a) si $\Gamma = \sqrt{\lambda}I$, encontraoms


$$J(x) = \frac12 \| Ax - b \| + \lambda \| x \|^2 .$$

Alternativamente existen los **mulitiplicadores de Lagrange**.

Es decir podemos formular el problema como

\begin{eqnarray}
\min_x \frac12 \| Ax - b \|^2  \\
\text{sujeto a la restriccion  } \| x \|^2 \le c
\end{eqnarray}

# Pseudoinversa
Que es la pseudoinversa? Para entender vamos a proponer la solucion de $Ax = b$.

Hay varios casos:

* Si $b \in \mathcal{R}(A)$, el sistema tiene solucion.

    * Si $\mathcal{N}(A) = \{ 0 \}$ la solucino es unica. La matriz $A$ es invertible y la solucion es:

# Algoritmo para la pseudoinversa.
Usamos SVD. Recuerde

$$ A = U \Sigma V^* .$$

Recuerde que

$$ \Sigma = \begin{pmatrix} \Sigma_r &|& 0 \\ --- & --- &--- \\ 0 & | & 0   \end{pmatrix} .$$

Como $U,V$ son ortogonales $U^{-1}=U^*$, $V^{-1}=V^*$,
entonces

quisiermos algo como

$$A^+ = V \Sigma^+ U^* .$$

Quien seira $\Sigma^+$? Si $\Sigma$ es $m \times n$,
$\Sigma^+$ es $n \times m$, ahora bien
$\Sigma^+$ tiene los inversos de los valores singulares cuando estos no sean 0, y los ceros de $\Sigma^*$.

De forma explicita

\begin{eqnarray*}
  \Sigma^+ = \begin{pmatrix}
    \sigma_1^{-1} & 0 & \cdots & \cdots & \cdots & \cdots & 0 \\
    0 &  \sigma_2^{-1} & \ddots &  &  &  & \vdots \\
    \vdots & \ddots &  \ddots & \ddots &  &  & \vdots \\
    \vdots &  & \ddots & \sigma_r^{-1} & \ddots &  & \vdots  \\
    \vdots &  &  & \ddots &  0 & \ddots  & \vdots  \\
    \vdots &  &  &  &  \ddots & \ddots   & \vdots \\
    \vdots &  &  &  &  & \ddots  & \vdots \\
    \vdots &  &  &   &  &  & \vdots \\
    0 & \cdots & \cdots & \cdots & \cdots & \cdots & 0 \\
  \end{pmatrix} \,.
\end{eqnarray*}


Explicitamente como equacion

\begin{eqnarray}
\sigma_{ij}^+ = \left \{
    \begin{array}{cc} \delta_{ij} \sigma_i^{-1} & i \le r \\ 0 & i>r  \end{array}  
\right .
\end{eqnarray}
Tenemos $A^+ = V \Sigma^+ U^*$.


$$x^+ = A^+ b = V \Sigma^+ U^* b .$$
