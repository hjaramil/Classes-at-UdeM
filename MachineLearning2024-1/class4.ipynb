# Autovalores/Autovectores

**Definicion:** Sea $A \in \mathbb{C}^{n \times n}$ si existe un vector $u \in \mathbb{C}^n$, $u \ne 0$ y $\lambda \in \mathbb{C}$
tales que

$$ A u = \lambda u  \tag{1}.$$
Vamos a ver metodos para hallar $u, \lambda$.

De la ecuacion (1) pasando a restar a la izquierda


$$ A u - \lambda u = Au - \lambda Iu = (A - \lambda I) u = 0 .$$

De la teoria de determinantes tenemos que si $Bx = 0$ y $x \ne 0$ entonces $\det (B) = 0$.
De forma que tenemos

$$ p(\lambda) = \det(A - \lambda I) = 0 .$$
Aca $p(\lambda)$ es un polinomio de grado $n$.
A este polinomio se le conoce como el **polinomio caracteristico**.


El algoritmo para hallar los autovalores/vectores es el siguiente.

1. Se resuelve (por $\lambda$) el polinomio caracteristico $p(\lambda) = 0.$
2. Una vez hallado $\lambda$ se resuelve (por $u$) la ecuacion (1)
$$ (A - \lambda I)u = 0 .$$
La solucion al punto 2 es un numero infinito de soluciones pues el determinante de la matriz acompanante es 0.

**Ejemplos:**
Considere la matriz

$$A = \begin{pmatrix} -5 & -3 \\ -6 & 2 \end{pmatrix}, $$
halle los autovalores y autovectores.


Usamos el algoritmo de arriba.
El polinomio caracteristico de $A$ es


\begin{eqnarray}
p(\lambda) &=& \det ( A - \lambda I) \\
&=& \det \begin{pmatrix} -5 - \lambda & - 3 \\ -6 & 2 - \lambda \end{pmatrix} \\
&=&  (-5 - \lambda)(2 - \lambda) - 18 \\
&=& -10 + 3 \lambda + \lambda^2 - 18 \\
&=& \lambda^2 + 3 \lambda - 28 \\
&=& (\lambda + 7 ) ( \lambda  - 4 ) = 0
\end{eqnarray}
Con esto  tenemos dos valores propios que son $\lambda_1 = -7$,
$\lambda_2 = 4$.

Para hallar $u$ correspondiente a $\lambda_1 = -7$, tenemos la ecuacion.


\begin{eqnarray*}
    \left \{ \begin{pmatrix}
      -5  & - 3 \\
      -6 & 2
    \end{pmatrix} -  \begin{pmatrix}
      -7  & 0 \\
      0 & -7
    \end{pmatrix}  \right \} \begin{pmatrix}
      x  \\
      y
    \end{pmatrix}
    &=&
    \begin{pmatrix}
      0 \\
      0
    \end{pmatrix}
    \\
    \begin{pmatrix}
      2  & - 3 \\
      -6 & 9
    \end{pmatrix}  \begin{pmatrix}
      x  \\
      y
    \end{pmatrix}
    &=& \begin{pmatrix}
      0 \\
      0
    \end{pmatrix}  .
  \end{eqnarray*}









Tomamos la primera ecuacion

$$ 2 x - 3y = 0.$$

Una solucion facil es $x=3$, $y=2$, entonces tenemos

$$ u = \binom{3}{2} .$$
Este no es el unico. Muchos paquetes muestran el vector normalizado, es decir este divido por $\sqrt{3^2 + 2^2}=\sqrt{13}.$

Es posible que un autovalor se repita y genera autovectores de multiplicidad 2 o mas. Esto significa que la matriz se reduce 2 o mas dimensiones. El siguiente ejemplo es the ChatGPT.

\[ A = \begin{pmatrix}
3 & 0 & 0 \\
0 & 3 & 1 \\
0 & 0 & 3 \\
\end{pmatrix} \]

For $\[ \lambda = 3 \]$, the system $\( (A - 3I) \mathbf{x} = \mathbf{0} \)$ leads to:

\[ \begin{pmatrix}
0 & 0 & 0 \\
0 & 0 & 1 \\
0 & 0 & 0 \\
\end{pmatrix} \begin{pmatrix}
x_1 \\
x_2 \\
x_3 \\
\end{pmatrix} = \begin{pmatrix}
0 \\
0 \\
0 \\
\end{pmatrix} \]

This simplifies to \( x_2 = 0 \), which means the eigenvector corresponding to \( \lambda = 3 \) is any vector of the form \( \begin{pmatrix} x_1 \\ 0 \\ x_3 \end{pmatrix} \), where \( x_1 \) and \( x_3 \) can be any non-zero real numbers.


**Ejemplo**:
Considere la matriz

$$ A = \begin{pmatrix} 1 & 1/2 \\ 1/2 & 1  \end{pmatrix} .$$
Halle los autovalres/vectores y explique lo que esta matriz le hace a un cuadrado con vertices $(1,1), (-1,1) (-1,-1), (1,-1)$.


Comencemos hallando los autovalores.

\begin{eqnarray}
\det ( A - \lambda I) &=& \det \begin{pmatrix} 1 - \lambda & 1/2 \\ 1/2 & 1 - \lambda \end{pmatrix} \\
&=& (1-\lambda)^2 - \frac14 \\
&=& (1 - \lambda - 1/2)(1 - \lambda + 1/2) \\
&=& (1/2-\lambda)(3/2 - \lambda).
\end{eqnarray}
Los autovalores son $\lambda_1 = 3/2$, $\lambda_2 = 1/2$.


Autovectores. Resolvemos la primera ecuacion del sistema $(A - \lambda I)u = 0$.

La primera ecuacion

$$ (1 - \lambda) x + \frac{y}{2} u =  0.$$
Como $\lambda_1=3/2$


$$ -\frac12 x + \frac12 y = 0.$$

Una solucion facil es $x=y$. De forma por ejemplo $x=y=1$ generan el autovector

$$ u = \binom{1}{1} .$$

De igual forma, repitiendo el proceso para $\lambda_2=1/2$ encontramos.

$$\frac12 x + \frac12 y = 0 .$$
La solucion general es $x=-y$, entonces un autovector $v$ es

$$ v = \binom{1}{-1}.$$

Aplicamos la matriz $A$ al cuadrado de arriba.
\begin{eqnarray}
\begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix}
\binom{1}{1} = \binom{3/2}{3/2}  \\
\begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix}
\binom{-1}{1} = \binom{-1/2}{1/2}  \\
\begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix}
\binom{-1}{-1} = \binom{-3/2}{-3/2}  \\
\begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix}
\binom{1}{-1} = \binom{1/2}{-1/2}  \\
\end{eqnarray}

Recuerde

$$ u = \binom{1}{1} \quad v = \binom{1}{-1} .$$

Tenemos

$$  \| u \| = \| v \| = \sqrt{1^2 + 1^2} = \sqrt{2}.$$

La salida de $Au=u'$ la salida de $Av = v'$.

Tenemos las siguientes magnitudes

$$  \| u' \| = \sqrt{ (3/2)^2 + (3/2)^2} = \sqrt{9/2} = 3/\sqrt{2}.$$

$$  \| v' \| = \sqrt{ (1/2)^2 + (1/2)^2} = \sqrt{1/2} = 1/\sqrt{2}.$$

Vemos como se **interpretan** los autovalores con este ejemplo.

En general si $\lambda_1, \lambda_2$ son los autovalores


$$ \frac{ \| u ' \|}{ \| u \|}  =\frac{ \| A u \|}{ \| u \|} = \frac{\| \lambda u \| }{ \| u \|} = \frac{\lambda \| u \|}{\| u \|} = \lambda .$$

En el caso que nos atanhe.


\begin{eqnarray}
\lambda_1 &=& \frac{\| u' \|}{\| u \|} = 3/2 \\
\lambda_2 &=& \frac{\| v' \|}{\| v \|} = 1/2 \\
\end{eqnarray}
