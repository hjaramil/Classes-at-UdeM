Dentro de ML existe los siguientes metodos.

* Metodos supervisados: Son metodos donde se tienen datos con etiquetas. Las etiquetas son las repuestas.
* Metodos no supervisados (unsupervised). No tienen etiquetas.
* Semisupervisados, algunos tienen etiquetas y otros no. Los sistemas de recomendacion funcionan asi. Por ejemplo en Netflix. Usted puede o no calificar la pelicula que esta viendo. Si la califica la esta metiendo en la supervision, si no no es supervisado.

* Metodos re-enforzados (reinforcement learning). Es por recompenzas. Como los juegos. Inicialmente existe un ambiente (environment) y un agente (el jugador) . Funciona por "rewards" (recompenzas) y "penalties" (penalidades). Este metodo no lo estudiamos en este curso.
