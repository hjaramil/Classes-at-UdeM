### K-means

Hay situaciones donde el numero de clases es facil. Por ejemplo
cuando se clasifica por genero (M-F), o tallas de ropa
S-M-L  (small, medium, large)

Dentro del analisis de agrupamiento hay varios algoritmos.

* k-means (algoritmo de medias en k clases)
* Density based (basados en densidad) (no lo vemos aca)
* HDBC : Hierarchical Density Based Clustering. Basado en Jerarquias.

En este curso solo vemos el k-means, que tiene limitaciones pero los otros metodos (mas sofisticas) se dejan como complemento de lectura.

Explico el algoritmo de k-means. Es un algoritmo geometrico.

Se proyecta tener $k$ grupos (como elejir k? buena pregunta)
Cada grupo tiene un centroide (de alli la palabra "mean")
Los $m$ puntos (datos) se van a asignar a cada cluster mediante
un vector $c$ de centros. Con mas detalle.

1. Decida cuantos grupos debe usar. Claramente $k < m$.
Esta pregunta se estudia mas adelante.

2. Elija $k$ centros aleatoriamente. $u_j$, $j=1,2, \cdots, k$.
Como son aleatorios no son centro de nada **todavia**.

,1,3. Inicialice un vector $c$ de $m$ componentes. Asigne a cada componente $c_i$, el indice del "centro" ($\mu_i$) mas cercano.

4.  Si el punto $x_i$ es mas cercano a $\mu_{c_i}$. Asignamos $c_i$ a centro. $c_i=1, 2 , \cdots k$. Esto nos divide los $m$ puntos en $k$ grupos. Veamos un ejemplo especifico. Digamos que el vector $c$

* $$ c =[1,1,2,3,1,3,2,2]$$. Esto quiere decir que

* $x_1, x_2, x_5$ todos estan mas cera al centro $\mu_1$.

* $x_3, x_7, x_8$ todos estan mas cera al centro $\mu_2$.

* $x_4, x_6$ estan mas cerca a $\mu_3$.

5. Calculamos nuevos centros. En este ejemplo

\begin{eqnarray}
\mu_1 &=& \frac{x_1 + x_2 + x_5}{3} \\
\mu_2 &=& \frac{x_3 + x_7 + x_8}{3} \\
\mu_3 &=& \frac{x_4 + x_6}{2}
\end{eqnarray}

6. Iteramos desde el paso 4.


La funcion de costo asociada a este problema es:


$$J(x, c, \mu ) = \sum_{i=1}^m \| x_i - \mu_{c_i} \|^2 \tag{1} .$$

Pregunta Daren:
Que garantiza que dos centros no converjan al mismo punto?
Si de entrada el algoritmo aleatorio junta dos puntos. Hay problema.
El algoritmo necesitaria algo asi como (tres clasters)

$\| \mu_i - \mu_j \| > 0$, $i \ne j$.

Una vez los centros iniciales no coincidan en ninguna parte, es dificil que
se fusionen en el proceso. Por lo menos para casos donde los datos tiene clara separacion.

El metodo de k-means asume una particion. Es decir, que no hay un elemento en mas de un  grupo. Que cada elemento esta asignado a **uno y solo un** grupo.

En la practica hay muchos casos que no son faciles. Vamos ver ejemplos

Un principio que deseariamos tener es

1. Que los objetos en el **mismo** cluster sean tan similares (cercanos) como sea posible. **Cohesion**. Mas de esto en la proxima clase.
