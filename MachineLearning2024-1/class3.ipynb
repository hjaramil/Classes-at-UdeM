Vamos a probar (basados en lo anterior) que si una terna satisface el teorema de pitagoas entonces es una terna correspondiente a un triangulo rectangulo.

Asumimos que el triangulo es rectangulo (se cumple el teorema de Pitagoras), vamos a probar que los catetos son ortogonales ($\langle x, y \rangle = 0$).

Tenemos:

$$ \|x \|^2 + \| y \|^2 = \| x + y \|^2 .$$

Entonces:

Asumiendo que los vectores son reales se tiene la conmutatividad del producto interno. Es decir $\langle x, y \rangle = \langle y, x \rangle $.
$$ \| x  + y \|^2 = \langle x+y, x+y \rangle = \langle x, x \rangle + \langle y, x \rangle + \langle x, y \rangle + \langle y, y \rangle = \| x \|^2 + 2 \langle x, y \rangle + \| y \|^2 .$$



De forma $\langle x, y \rangle = 0.$

Al reves tambien pasa. si $\langle x, y \rangle = 0$, entonces (tarea) debemos probar el teorema de Pitagoras $\| x + y \|^2 = \|x\|^2 + \| y \|^2.$

**Definicion: Matrix ortogonal**: Una matriz cuadrada es ortogonal si sus columnas son mutuamente ortogonales y ademas son de norma 1. La palabra ortogonal implica que las comunas son normalizadas. Yo pensaria que la definicion deberia ser **ortonormal**.  Pero no.

Mas concretamente, si $a_i, a_j$ son dos columnas de la matriz $A$
entonces $\langle a_i , a_j \rangle = 1$ si $i=j$, o 0 si $i \ne j$. Aprovechamos y definimos el delta de Kronecker.


\begin{eqnarray}
\delta_{ij} =  \left \{
    \begin{array}{cc}
    1 & \text{si  } i = j \\
    0 & \text{si  } i \ne j
    \end{array}
    \right .
\end{eqnarray}

Con esto tambien podemos definir la matriz identidad. $I$.
en el espacio $\mathbb{R}^{n \times n}$.

$$ I = I_n = \begin{pmatrix} 1 & 0 & 0 & \cdots & 0 \\
0 & 1 & 0 & \cdots & \vdots  \\
\vdots & \vdots & \vdots & \ddots & 0  \\
0 & \cdots & \cdots & 0 & 1
\end{pmatrix} . $$

Usando la definicion del delta de Kronecer escribimos
$I= (\delta_{ij})$, $i=1, \cdots, n$, $j=1, \cdots , n$, .

Un ejercicio para el estudiante es probar que
$AI = IA = A$ para cualquier matriz $A \in \mathbb{C}^{n \times n}$.

Teorema: Si $A$ es ortogonal, entonces $A^* A = I$.

Prueba: Llamemos a los vectores columna de $A$, $v_i$.
La entrada $ij$ de la matriz $A^*A$ esta dada
$v_i^* v_j$ pero como $v_i, v_j$ son ortogonales entonces

$$ v_i^* v_j =  \langle v_i, v_j \rangle = \delta_{ij}.$$
Es decir $A^* A = I$.

Ejemplo:
Sea
$$ A= \begin{pmatrix} \mathrm{i}  & 0  \\ 0 & \mathrm{-i} \end{pmatrix} .$$

Entonces
$$ A^* = \begin{pmatrix} -\mathrm{i}  & 0  \\ 0 & \mathrm{i} \end{pmatrix} .$$

Ahora

$$ A^* A  = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} = I_2.$$

**Definicion: Inversa de una matriz :** Sea $A$ una matriz cuadrada de orden $n$, si existe $B$ tal que $AB=BA=I$ decimos
que $B$ es la inversa de $A$ y escribimos $B=A^{-1}$.
Podemos hablar de $A/B$ como $A B^{-1}$.
Una matriz que tiene inversa se llama **no singular**, si no
tiene inversa se llama **singular**.

Las matrices ortogonales tiene una propiedad importante. Es que
mapean vectores a vectores con la misma norma. Preservan la norma, son **isometricas**.


**Teorema:** Las matrices ortogonales son isometricas. Es decir
$ \| Ax \| = \| x \|$.



Prueba:

$$\| A x \|^2 = \langle Ax, Ax \rangle = (Ax)^* Ax = x^* A^* A x = x^* I x = x^* x = \| x \|^2. $$
Sacando raiz cuadrada a ambos lados


$$ \| Ax \| = \| x \| .$$

Probemos usando el Teorema de arriba (determinates) que $A^* A  = A A^*$.


**Teorema**: Asumas $AB=I$, si $A,B$ son matrices cuadradas, entonces $BA=I$.

Prueba: Si $AB = I$,

$$ B = B(AB)= (BA) B.  \tag{1} $$
Como el determinante de $I$ es 1 ( $\det I  = 1$).
De forma que $\det(AB) = \det(A) \det(B)=1$, entonces
$\det(A) \ne 0, \det(B) \ne 0$ de forma que $B$ es invertible, o existe $B^{-1}$.  

En la ecuacion (1) multiplicamos a ambos lados por $B^{-1}$ a derecha.

$$ I = (BA) .$$
Es decir $AB=BA$ y concluimos que una matriz y su inversa conmutan. De forma tambien tenemos que $A^* A = A A^* = I$.

**Teorema: Inverso del producto de matrices**: Si $A, B$ son
matrices no singulares, entonces $(AB)^{-1}=B^{-1} A^{-1}$.

Veamos la prueba


$$(B^{-1} A^{-1})(AB) = B^{-1}(A^{-1} A) B = B^{-1} B = I .$$

Definamos espacio rango y espacio nulo de una matriz.


**Definicion:** Sea $A$ una matriz $m \times n$, definimos
el **rango** de la matriz como el espacio generado por las columnas de la matriz. Lo notamos como $\mathcal{R}(A)$.
Mas concretamente, si las columnas de la matriz $A$ estan dadas por los vectores $A_i$, el rango de la matriz $A$ esta dado por


$$\mathcal{R}(A) = \sum_{i=1}^n \alpha_i A_i .$$
donde $\alpha_i$ es un complejo arbitrario.


El espacio **nulo** (**kernel**) de la matriz $A$ es el espacio de todos los vectores $x$ tales que $Ax=0$.


$$\mathcal{N}(A) = \{ x : A x = 0\} .$$





### Autovalores/Autovectores (Eigenvalues/Eigenvectors)
Nocion: Asumamos espacios reales. Los autovectores son vectores que no cambian de direccion cuando se les aplica la matriz $A$,
los autovectores son escalares de "estiramiento" de los autovectores en la operacion $Ax$. Mas concremaente si

$$ Ax = \lambda x .$$
decimos que $x$ es una **autovector** de la matriz $A$ y que
$\lambda$ es un **autovalor** de la matriz $A$.
