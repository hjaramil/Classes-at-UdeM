**Definicion 2.5.1**(Derivada de una matriz): Asuma $A = A(x)$, $m \times n$, con $x \in \mathbb{R}^p$, entonces $A=[a_{ij}]$

$$\frac{\partial A}{\partial x_k} = \left [  \frac{\partial a_{ij}}{\partial x_k}  \right ] $$

**Teorema 2.5.1**: El gradiente de $A$ para $A$ constante $a_{ij}$ con constantes es 0.

$$\frac{\partial A}{\partial x_k} = \left [  \frac{\partial a_{ij}}{\partial x_k}  \right ] = \left [  0  \right ] $$
pues $a_{ij}$ es constante y de calculo diferencial $\partial a_{ij} /\partial x_k = 0$.

Como el calculo $d (ax)/dx=a$ $x,a  \mathbb{R}$, en matrices paasa algo parecido

**Teorema 2.5.2** Si $A$ es un matriz $m \times n$ con $X \in \mathbb{R}^n$, entonces si $A$ es constate

$$\nabla (A x) = A $$
Veamos, tomemos la componente $k$ de la derivada

$$\frac{\partial Ax}{\partial x_k}  $$
$$ (Ax)_{i}  = \sum_{j=1}^n a_{ij} x_j  $$
$$ \frac{(Ax)_{i}}{\partial x_k}   = \sum_{j=1}^n a_{ij} \frac{\partial x_j}{\partial x_k} = \sum_{j=1}^n a_{ij} \delta_{jk} = a_{ik}   $$

Tenemos que

$$\frac{ A x }{\partial x_k} = \nabla_K (A x) = A_k $$
donde $A_k$ es la $k$ columna de $A$.

**Teorema 2.5.3** (derivada del producto interno): Sean $f(x)$ y $g(x)$ dos funciones vectoriales definidas en el dominio $x \mathbb{R}^n$.

Entonces

$$\frac{\partial \langle f, g \rangle }{\partial x_i}  =
\left \langle   \frac{\partial f}{\partial x_i}, g \right \rangle +
\left \langle   f  , \frac{\partial g}{\partial x_i} \right \rangle
$$

**Prueba**
Recordemos que


$$ \langle f, g \rangle = \sum_{j=1}^n \overline{f_j(x)} g_j(x)  $$

Entonces

\begin{eqnarray}
\frac{ \partial \langle f, g \rangle }{\partial x_i } &=&  \sum_{j=1}^n \frac{\partial \overline{f_j(x)}}{\partial x_i} g_j(x) + \overline{ f_j(x)} \frac{\partial g_j(x)}{\partial x_i} \\
\end{eqnarray}
Por definicion de producto interno, tenemos la suma de dos productos internos.


$$\frac{\partial \langle f, g \rangle }{\partial x_i}  =
\left \langle   \frac{\partial f}{\partial x_i}, g \right \rangle +
\left \langle   f  , \frac{\partial g}{\partial x_i} \right \rangle
$$


A veces escribrimos
$$\nabla \langle f, g \rangle = \langle \nabla f, g \rangle + \langle f, \nabla g \rangle  $$
El gradiente $\nabla f$ es una matrix y $f,g$ son vectores entoncse cuidado con esta notacion. Podemos simplemente escoger la componente $k$ del gradiente (columna)

$$\nabla_k \langle f, g \rangle = \langle \nabla_k f, g \rangle + \langle f, \nabla_k g \rangle  $$

### Problema de minimios cuadrados.
Queremos resolver $Ax = b$, pero hay infinitas soluciones. Entonce la menor de
 $\| A x - b \|^2$  es la solucion de minimos cuadrados.

 **Ejemplo 2.5.1** (minimos cuadrados): Sea
 $$f(x) = \frac12 \| Ax  - b \|^2 $$
 $x  \in \mathbb{R}^n$, $A$ esa $m \times n$. Encuentre el gradiente $\nabla f$.

 **Solucion**:

 $$f(x) = \frac12 \langle Ax -b, Ax - b \rangle  $$
 Usamos todas las propiedades de arriab.

 \begin{eqnarray}
 \nabla_k f(x) &=& \frac12 [ \langle \nabla_k (Ax -b) , Ax-b \rangle + \langle Ax -b , \nabla_k (Ax - b) ] \\
 &=& \langle \nabla_k ( Ax - b), Ax-b \rangle \quad, \quad \text{el producto interno conmuta en los reales} \\
 &=& (A_k, A x- b \rangle
 \end{eqnarray}
 para $k=1,2, \cdots, p$
 Matricialmente esto se escribe

 $$\nabla f(x) = A^* ( Ax - b)  $$
 Si queremos opitimizar (minimo) el gradiente debe ser 0, pero esto implica

 $$\nabla f(x) = A^* ( Ax - b) = 0 $$
 $$  A^* Ax  = A b $$
 $$  A^T Ax  = A b $$

 ## Regularizacion de minimios cuadrados.

 Veamos por etapas. El problema que queremos resolver es $Ax=b$.
 * Si $A$ es invertible entonces
 $$x = A^{-1} b  $$
 y listo.
 Pero en la practica $A$ ni siquiera es cuadrada. $m \ne n$
 Veamos dos casos con $m \ne m$
    * Si $m > n$ y el numero de filas de $A$ es linealmente independiente.
    El sistema es **sobredeterminado**. El espacio nulo e $\{0\}$ pero el rango de $A$ no cubre el espacio.

    * Si $m < n$. El sistema es sub-determinado (underdetermined).
    En este caso el rango cubre todo el espacio y el espacio nulo es mayor que el trivial.

**Funciones convexas**: Son las mejores para minimiar. La funcion convexa se define como una done, si dos puntos en la grafica se unen, la grafica esta totalmente debajo de ella. Si la funcion es real y doblemente diferenciable, esta es convexa si su segunda derivada es positva. La funcion $f$ del ejemplo
(es una paraboloide) y si es estrictamente convexa tiene un minimo unico.

El problema $f(x) = \frac12 \| Ax - b \|^2$ no necesariamente tiene minimo.
Basta con las columnas de $A$ no sean linealmente independientes. Entonces, Tokhonov llego con la idea de poner una funcion **penalty**


$$f(x) = \| A x - b \|^2 + \| \Gamma x \|^2  $$

Si tomamos $\nabla f = 0$ como en ejemplo de arriba.

$$[(A^T A)  + \Gamma^T \Gamma) ] A^T b = A^T b  $$
$\Gamma$ es positiva definida, $A^T A$ es no-negativa definida (o sea que todos sus autovalores son $\lambda_i \ge 0$ y la suma (tarea) es positiva definida.

Es decir $A^T A + \Gamma^T \Gamma$ es **postiva definida** y por tanto invertible  y

$$ x = (A^T A + \Gamma^T \Gamma)^{-1} A^T b . $$
$x$ en la ecuacion anterior es la solucion regularizada de minimos cuadrados.

Es commun ver que $\Gamma = \sqrt{\lambda} I$.  Entonces

$$ x = (A^T A +  \lambda I)^{-1} b . \tag{1} $$
A este tipo de inversion en estadistica se le conoce como **ridge regression**.

Una forma heuristica de llegar a la Ecuacion (1) es:

* Comience con el problema original $Ax=b$.
* Premultiplique por $A^T$: $A^T A x = A^T  b$ (estas son las ecuaciones normales).
* Si el espectro de $A^T A$ tiene valores muy cercanos a 0, o 0, entonces le hacemos un " corrimiento" sumando $\lambda I$ y resolviendo el sistema que queda.
$$x  = (A^T A + \lambda I)^{-1} b . $$

Otra forma de entender esto es con multiplicadores de Lagrange.

Piense en una funcion de costo

$$J(x) = \frac12 \| A x - b \|^2 + \lambda \| x \|^2 $$

Si escribimos esto pensando en multiplicadores de Lagrange

\begin{eqnarray} \min_x \frac12 \| Ax - b \|^2  \\
\text{ Sujeto a } \| x \|^2 \le c,  c > 0.
\end{eqnarray}
