Sea $A$ una matriz $m \times n$ el **range** (rango, espacio columna) es el generado por las columnas de $A$. Es decir si $A_i$ es una columna de $A$ el
rango $\mathcal{R}(A)= \sum_{i=1}^n \alpha_i X_i$, $\alpha_i$ es escalar.
El espacio nulo es el espacio que anula los vectores en $n$ dimensiones. Es decir,

$$\mathcal{N}(A) = \{ x \quad | \quad A x = 0   \}  $$



# Eigenvalues/vectors (autovalores/autovectores, valores singulares, vectores singulares)
**Definicion 2.2.1** (autovalor/autovector). En ingles Eigen(vector/value), el "eigen" es del Aleman, (auto, propio). Dada una matriz $A$ cuadrada ($n \times n)$ , si existe un $\lambda \in \mathbb{C}$ tal que

$$Ax = \lambda x \tag{1} $$
para $x \ne 0$. Decimos que $\lambda$ es una valor propio de $A$ y que $x$ es el vector propio correspondiente a $\lambda$.
[Wikipedia](https://en.wikipedia.org/wiki/Eigenvalues_and_eigenvectors#/media/File:Eigenvectors.gif) ilustra el concepto dinamicamente.

Note que la ecuacion $Ax = \lambda x$ me dice que el operador $A$ esta "mandando" el vector $x$ a un multiplo.

## Como se hallan? algoritmo.
De la ecuaci'on (1)

$$Ax - \lambda I x = 0 $$
$$(A - \lambda I) x = 0 $$
Recuerden que $x \ne 0$. Entonces, o sea que $A - \lambda I$ es **singular** y

$$p(\lambda) = \det (A - \lambda I) = 0. \tag{2} $$
Aca $p(\lambda)$ es un polinomio de orden $n$


**Ejemplo 2.2.2**: Encuentre los auto-valores/vectores de la matriz

$$A = \begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix} $$

Segun el algoritmo (2).

\begin{eqnarray}
\det ( A - \lambda I) &=& \det \begin{pmatrix} 1- \lambda & 1/2 \\ 1/2 & 1 - \lambda\end{pmatrix}  \\
&=& (1 - \lambda)^2 - \frac14 \\
&=&\left (1 - \lambda - \frac12 \right ) \left (1 - \lambda + \frac12 \right ) \\
&=& (1/2 - \lambda) ( 3/2 - \lambda) = 0.
\end{eqnarray}

La solucion es $\lambda_1 = 3/2$  y $\lambda_2 = 1/2$.

### Ejemplo en `Python`

```python
import numpy as np
import matplotlib.pyplot as plt

# generate points in a unit circle
theta = np.arange(0, 2*np.pi, np.pi/24)
circle = np.array([np.cos(theta), np.sin(theta)])

# definamos la transformacion
A = np.array([[1, 0.5],[0.5,1]])

# aplicamos el Mapping
transformed = A @ circle

# origen
origin_x = np.zeros(circle.shape[1])
origin_y = np.zeros(circle.shape[1])

# grafica
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect('equal')

# draw original vectors in black
ax.quiver(origin_x, origin_y, circle[0], circle[1], color='black', label='original vectors',
          scale_units='xy', angles='xy', scale=1, alpha=0.6)

# draw mapped vectors
ax.quiver(origin_x, origin_y, transformed[0], transformed[1], color='red', label='mapped vectors',
          scale_units='xy', angles='xy', scale=1, alpha=0.6)

# anotar autovectores
ax.text(1.4, 0.0, '$\\vec{u}$', fontsize = 12)
ax.text(0.0, 1.1, '$\\vec{v}$', fontsize = 12)

# Axis
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

ax.grid(True)
ax.title("autovectores...")

plt.tight_layout()
plt.show()


```

### Teorema espectral para operadores (matrices) autoadjuntos.

***Teorema 2.2.2**: Si $A$ es autoadjunta, entonces los autovalores son reales y dos autovectores $u,v$ correspondientes a autovalores diferences son ortogonales.

**Prueba**: Sin perdida de generalidad podemos asumir autovectores $\| u_j\|=1$. Estudamos las **formas cuadraticas** $\langle A u_j, u_u \rangle$.

Probemos un lemita:
$$\langle A x, y \rangle = (Ax)^* y = x^* A^* y = \langle x, A^* y\rangle \tag{3} $$

1.
$$\langle A u_j, u_j \rangle = \langle \lambda_j u_j, u_j \rangle =
\overline{\lambda_j} \langle u_j, u_m \rangle = \overline{\lambda_j} \| u_j \|^2 = \overline{\lambda_j} \tag{3} $$

Ahora cmo $A$ es auto-adjunto
$$\langle A u_j, u_j \rangle = \langle u_j, A^* u_j \rangle = \langle u_j, A u_j \rangle  = \langle u_j, \lambda_j u_j \rangle = \lambda_j \| u_j \|^2 = \lambda_j \tag{4} $$

De (4) = (3)
$$\overline{\lambda_j} = \lambda_j $$
Otro lemita. Si un numero es igual a su conjugado entonces es real.
$$a+ \mathrm{i} b = \overline{a + \mathrm{i} b } = a - \mathrm{i} b  $$
Restando $2 \mathrm{i} b  = 0$. $b=0$.
O sea que los autovalores son **reales**.

2. Asumamos dos autovalores distintos $\lambda_1 \ne \lambda_2$
Ahora

$$\langle Au, v \rangle = \langle \lambda_1 u , v \rangle = \lambda_1 \langle u, v \rangle . $$

Ahora
$$\langle A u, v \rangle = \langle u, A v \rangle = \langle u, \lambda_2 v \rangle = \lambda_2 \langle u , v \rangle  $$


Restando las dos anteriores

$$(\lambda_1 - \lambda_2) \langle u , v \rangle = 0 $$
Como $\lambda_1 \ne \lambda_2$

$\langle u, v \rangle = 0$.

Pensemos en el espacio $\mathbb{C}^n$ sabemos tenemos una base ortonormal
$\{ x_i \} , i=1,2, \cdots ,n $
ademas $x_i x_j = \delta_{ij}$
$$ x_i^* A x_j = x_i^* \lambda_j x_j = \lambda_j \delta_{ij}  $$

Matricialmente escribimos

$$Q^* A Q = \Lambda  $$
donde $\Lambda$ es una matriz diagonal con los autovalores en su diagonal.

De aca podemos multiplicar a izquierda por $Q$ y a derecha por $Q^*$

$$Q Q^* A Q Q^* = Q \Lambda Q^*  $$


$$A  = Q \Lambda Q^*  $$

Que pasa si hay autovalores repetidos.

### Distintas formas de multiplicar matrices especiales.
Sea $A$ una matrix $m \times n$, $B$ una matriz $n \times p$, $x$ un $n$-vector, $\Lambda$ una matriz diagonal (autovalores) $n \times p$
$\Lambda$ tiene ceros por fuera de la diagonal princial. Entonces.

1.
$$AB = \left [ AB_1 | A B_2 | \cdots | A B_p   \right ]  $$
donde $B_j$ es la $j$ columna de $B$.

2.

$$A x = \sum_{j=1}^n x_j A_j  $$
donde $A_j$ es la $j$ columna de $A$.

3. Si $p \ge n$

$$A \Lambda = \left [ \lambda_1 A_1 | \lambda_2 A_2 | \cdots | \lambda_n A_n | 0_{n+1} |  0_{n+2}  \cdots 0_{p} \right ]  $$
si $n < n$

$$A \Lambda = \left [ \lambda_1 A_1 | \lambda_2 A_2 | \cdots | \lambda_p A_p |  \right ]  $$
si $n < n$

4. Si $u_i$ son vectores columna de una matriz $m \times m$ $U$ y $b \in \mathbb{C}^m$
entonces

$$ U^* b = \begin{pmatrix}  u_1^* b \\ u_2^* b  \\ \vdots \\ u_m^* b \end{pmatrix}  $$


5. Block matrices (matrices en bloque).
