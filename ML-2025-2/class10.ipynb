## Sets (conjuntos)
Es un grupo de elementos que comparten cierta propiedad. Aca **no  hay repetidos**.

Supongamos que el orden importa

1. **Importa el orden** ( $k$-permutaciones). Vamos a usar el modelo de las cajas. Tenemos un valde con $n$ objetos distintos.
    * La primera vez sacamos un objeto y quedan $n-1$
    * La segunda vez sacamos un objeto y quedan $n-2$
    * $\vdots$
    * Para $k$ vez tenemos
$$\text{count} = n \times n-1 \times n-2 \times \cdots (n-k+1) =
\frac{n!}{(n-k)!} = n^{\underline{k}} $$

A esta expresion se le conoce como **falling factorial** (factorial que cae).
De cuantas formas puedo extraer $k$ objetos de un valde con $n$ objetos, sin repeticion pero importando el orden.

$$ k \le n$$
Ai $k=n$ entonces
$$\text{count} = n! $$


2. **No importa el orden** ($k$-combinacion):

En este caso, cada que saquemos $k$ objetos, se cuentan $k$ objetos
$k!$ veces.

$$\text{count} = \frac{n \times n-1 \times n-2 \times \cdots (n-k+1)}{k!} =
\frac{n!}{k! (n-k)!} = \binom{n}{k} $$

A esto se le llama el **coeficiente binomial** por razones que explico mas adelante. Tambi'en es el numero de subconjuntos de $k$ elementos que puede tomar de un conjunto de $n$ elementos.

## Muliconjuntos (multisets)
En este caso, por ejemplo

$$M = \{ a, a, a, b, b, c \} $$
Este es un multiset que tiene $a$ 3 veces , $b$ dos veces.

Ejemplos de multisets. La descomposicion prima de un numero.

M'as formalmente un mutiset es una funci'on en los naturales (0)

\begin{eqnarray}
m : A &\to& \mathbb{N} \\
x &\mapsto& m(x) = \text{cuantas veces esta el objeto en $A$ }
\end{eqnarray}

Lo mismo que en conjuntos consideramos dos casos:

1. Importa el orden ( permutaciones).
Veamos el ejemplo de la tabla de la palabra bookkeeper.

La pregunta: Cuantas palabras distintas podemos escribir con las letras de
la palabra "bookkeeper" .
Tenmos 10 letras. Entonce s esas 10 letras nos generan $10!$ permutaciones.
Pero, como por ejemplo la letra "e" esta repetida 3 veces entonces estamos contando $e_1e_2e_3, eee, eee, eee, eee ,eee$.
Estariamos contanto $3!=6$ veces mas si la letra "e" fuera distinta en cada caso.  Debemos entonces dividir por $3!$. Lo mismo con las demas. El conteo es

$$\text{count} = \frac{10!}{3! 2! 2!  1! 1! 1!} = 151200$$

Mas generalmente pensemos en lo siguiente, si $A$ es el multiset

$$n = n=\#(A, m)= m(e) + m(o) + m(k) + m(b) + m(p) + m(r)=3 + 2 + 2 + 1 + 1 + 1 = 10 $$

$m_1=3, m_2=2, m_3=2, m_4=1, m_5=1$ (multiplicidades)

$$\text{count} = \frac{n!}{m_1! m_2! \cdots m_k!} = \frac{m_1 + m_2 + \cdots + m_k}{m_1! m_2! \cdots m_k!} \tag{1} $$

Si $k=2$ esta es la misma binomial pues

$$\frac{m_1 + m_2}{m_1! m_2!} = \frac{n}{m_1! (n-m_1)!}= \binom{n}{m_1} = \binom{n}{m_2}  $$
Verifique esto.
Interesante que de dos formas distintas llegamos a la binomial:
1. no importa el orden y sin repeticion.
1. importa el orden pero con repeticion.

La ecuacion (1) se llama la **multinomial**. Veamos por que.

Piense en el problema

$$(x_1 + x_2 + \cdots + x_k)^n = \sum_{i_1 + i_2 + \cdots + i_k=n} c_{i_1 i_2 \cdots i_k} x_1^{i_1} x_2^{i_2} \cdots x_k^{i_k}  \tag{2} $$


En este problema el conjunto es (multiset)

$$A= \{ x_1, x_2, \cdots , x_k \} $$
y
$$m_j = m(x_j) = i_j $$
es la multiplicidad $i=1,2, \cdots, k$.

El coeficiente $c_{i_1 i_2 \cdots i_k}$ es

$$c_{i_1 i_2 \cdots i_k} = \frac{n!}{i_1! i_2! \cdots i_k!} = \binom{n}{i_1, i_2, \cdots, i_k} $$
Este coeficiente se llama el coeficiente binomial.

Aplicaciones intersantes.  En el caso $k=2$

$$(a + b)^n = \sum_{i=1}^n \binom{n}{i} a^i b^{n-i} $$
Esta el binomio de Newton.
Recuerden que $\binom{n}{k}$ cuenta el **number de subconjuntos de $k$ elementos de un conjunto de $n$ elementos**.
De forma que si $a=b=1$


$$2^n = \sum_{i=1}^n \binom{n}{i} $$
Esta formula me dice que el **numero total de subconjuntos de un conjunto con cardinal $n$ es $2^n$**

Este resultado se pudo haber obtenido de una forma mas simple.
Piense en un conjunto de $n$ elementos. Tome el modelo de cajas.
Cada caja tiene un elemento. Ejemplo, el conjunto $\{a,b,c,d \}$
Tomemos el subconjunto $a,b$

Ponga las cuatro cajas.
* En la primera marca 1 (corresponde a $a$)
* En la segunda marca 1 (corresponde a $b$)
* En la tercera marca 0 ( corresponde a $c$)
* En la cuarta marca 0 (corresponde a $d$).


* **No importa el orden**.  En este caso observe la ecuacion 2
y queremos contar el numero todal de terminos del multinomio.
O sea de cuantas formas podemos escibir

$$i_1 + i_2 + \cdots i_k=n $$
En palabras: De cuantas formas podemos resolver la Ecuacion Diofantina

$$i_1 + i_2 + \cdots i_k=n $$
donde $0  \le i_1, i_2, \cdots, i_n  \le n$

Otros problemas correspondientes a esta ecuacion diofantina

* Numero de combinaciones de $k$ objetos tomamos de $n$ objetos con **repeticion**.

* Numero de formas en que $n$ objetos identicos se pueden repartir entre $k$ cajas.

La solucion de este problema se da con un metodo que se llama
[stars and bars](https://en.wikipedia.org/wiki/Stars_and_bars_(combinatorics)).

Para explicar este metodo vamos usar un ejemplo.

**Ejemplo**: Cuantas soluciones tiene la ecuacion diofantina
$$x_1 + x_2 + x_3 = 2 $$
donde $x_i=0, 1, 2$

Podemos resolver el problema a mano

$$1 + 1 + 0 =  2 $$
$$1 + 0 + 1 =  2 $$
$$0 + 1 + 1 =  2 $$
$$2 + 0 + 0 =  2 $$
$$0 + 2 + 0 =  2 $$
$$0 + 0 + 2 =  2 $$

En vez de stars (estrellas) y barras usamos "+" y barras.

$$| + | + 0 =    $$
$$| + 0 + | =   $$
$$0 + | + | =   $$
$$|| + 0 + 0 =   $$
$$0 + || + 0 =   $$
$$0 + 0 + || =   $$

Quitando los ceros
$$| + | +      $$
$$| +  + |    $$
$$0 + | + |    $$
$$|| +  +     $$

$$0 +  + ||    $$

El problema se reduce a contar de cuantas formas puede
de 4 objetos tomar dos barras.
Por ejemplo, en el primer objeto las barras estan en
1,3
Por ejemplo, en el segundo objeto las barras estan en
1,4

Esto es equivalente a contar de cuantas formas podemos tomar 2 objetos de 4

$$\text{count} = \binom{4}{2} = \frac{4!}{2! 2!}= \frac{4 \times 3!}{4}=6 $$

En general pensemos en la ecuacion diofantina
$$x_1 + x_2 + \cdots + x_k = n $$
Si la llevamos a "barras" y "+" tendriamos elementos de la forma

$$|| + ||| + \cdots + || + | $$
En este caso seria $2 + 3 + \cdots + 2 + 1=n$.

Cuantos objetos hay: $n$ barras y $k-1$ "+"
El problema se reduce a contar cuantos grupos de $n$ elementos hay de un conjunto de $n+k-1$

Es decir

$$\text{count} = \binom{n+k-1}{n} = \binom{n+k-1}{k-1} \tag{3} $$

Es comun en la practica que la diofantina se escriba intercambiando $n$ con $k$

$$x_1 + x_2 + \cdots + x_n = k $$
En cuyo caso el conteo es

$$\left(\!\!\binom{n}{m}\!\!\right) =  \text{count} = \binom{n+k-1}{k} =
\frac{(n+k-1)!}{k! (n-1)!} = \frac{n(n+1)(n+2) \cdots (n+k-1)}{k!}= \frac{n^{\overline{k}}}{k!}  $$
A esta ultima expresion $n^{\overline{k}}$ se le conoce como el "raising factorial"

En resumen

$$\text{choose} = \binom{n}{k} = \frac{n^{\underline{k}}}{k!} $$
$$\text{mutlichoose} = \left(\!\!\binom{n}{k}\!\!\right) = \frac{n^{\overline{k}}}{k!} $$

# Redes Neuronales Artificiales (ANN: Artificial Neural Networks).
Vamos a ver dos razones importantes para la existencia de las ANN:

## Teorema de  la unviersalidad
[universality theorem](https://en.wikipedia.org/wiki/Universal_approximation_theorem).
El teorema dice que "casi" todas las funciones que usted use en la practica
$A \subset \mathbb{R}^n \to \mathbb{R}^m$ se puede aproximar tanto como quiera con una red neuronal, con profundidad $d=\max\{n+1, m \}$.

Recuerden que se pueden hacer muchas aproximaciones:
1. Fourier: Series de Fourier. Las bases son senos y cosenos (exponenciales complejas). Aproximan funciones con un numero finito de descontinuidades.
2. Weierstrass: Polinomios aproximan funciones continuas.
3. Taylor: Las bases son polinomios. Se aproximan funciones analiticas.

Aca las bases anteriores no aplican. Por que pueden tener oscilaciones muy rapidas (altras frecuencias) y puntos donde el gradiente se hace pequenho. Es importante que el gradiente no se haga muy pequenho. Esto genera problemas como lo vamos  a discutir mas adelante.

Entonces existen otras funciones (bases). Al final de la historia las redes neuronales construyen combinaciones lineales de las bases con coeficientes (pesos $\theta$) que son los que aprende el sistema. Se quieren funciones que sean monotonas. Por ejemplo la sigmoid ($S(x)=1/(1+\exp(-x))$.
La primera persona que probo esto es Cibenko.

## Historia
* En 1943 dos cientificos: Warren McCulloch (neurofisiologo) y el otro Walter Pitts (joven matematico) disenaron un modelo matematico que simula como funcionaria una neurona logica con circuitos electricos, simulando el "or", "and" ,etc.

* En 1958: Frank Rosenblatt, trabajando en el labrorio de Cornell Aeronautical Laboratory, desarrollo un modelo llamado **Perceptron**. ESte es el origen de las redes neuronales modernas.

* Bacpropagation: Este es alma detras de ANN. Este es el que encuentra el valor de los pesos (parametros) $\theta_i$. Se propuso en los 1960s, pero fue hasta los 1990s que se implement'o de forma exitosa.
