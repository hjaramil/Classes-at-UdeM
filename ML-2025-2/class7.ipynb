## Mas dimensiones:

$$h(\theta_0, \theta_1, \theta_2, x, y) = \theta_0 + \theta_1 x + \theta_2 y . $$
Evaluando esta ecuación en dos vectores $x,y$.

$$h(\theta_0, \theta_1, \theta_2, x_1, y_1) = \theta_0 + \theta_1 x_1 + \theta_2 y_1 . $$
$$h(\theta_0, \theta_1, \theta_2, x_2, y_2) = \theta_0 + \theta_1 x_2 + \theta_2 y_2 . $$
$$h(\theta_0, \theta_1, \theta_2, x_3, y_3) = \theta_0 + \theta_1 x_3 + \theta_2 y_3 . $$
$$\vdots $$

$$h(\theta_0, \theta_1, \theta_2, x_m, y_m) = \theta_0 + \theta_1 x_m + \theta_2 y_m . $$
En este caso

$$
\begin{pmatrix}
1 & x_1 & y_1 \\
1 & x_2 & y_2 \\
\vdots & \vdots & \vdots \\
1 & x_m & y_m \\
\end{pmatrix}
\begin{pmatrix}
\theta_0 \\
\theta_1 \\
\theta_2 \\
\end{pmatrix}
=
\begin{pmatrix}
y_1 \\
y_2 \\
\vdots \\
y
\end{pmatrix}
$$
Esto es **regresion lineal multivariada**.

El problema se formula como

$$X \theta = y \tag{1} $$
En general, para regresion podemos tener mas muestras que "features" (atributos) en este ejemplo solo hay 2 atributos.
entonces si premultiplicamos la Ecuacion (1) por $X^T $

$$X^T X \theta = X^T y $$
y $X^T X$ es no-singular (invertible) por que las columnas son linealmente independientes (pruebelo). Si dudamos entonces regularizamos

$$(X^T X + \lambda I)  \theta = X^T y $$
$\lambda>0$, $\lambda \ll 1$.  Este problema define una matriz
$$A = X^T X + \lambda I $$
que es positiva definida.
Sobre esta se recomienda $CG$.
La solucion analitica es

$$\theta = A^{-1}y $$
Si las dimensiones son muchas, no se deben resolver las ecuaciones normales de forma analitica. Se usan metodos de gradiente.

No hemos definido como normalizar los datos, pero usted pueede preguntar que al normalizar la matriz de datos $X$ estamos cambiando el problema. La respuesta es si pero no.

La ecuacion original es

$$X \theta = y \tag{1} $$
Si normalizamos $X$ a $X_n$
el problema es

$$X_n \theta_n = y  \tag{2} $$
Resolvemos el problema (2) es m'as estable.

Para normalizar se almacenan hiper-par'ametros en la memoria: por ejemplo media, desviaci'on estandard, etc (lo explicamos en un minuto).

Al resolver la Ecuacion 2. Tenemos $\theta_n$.
Que hacemos con esto?

No unos datos, por ejemplo la casa de area 2000 pies cuadrados y cuartos 5. Cual es el precio?

1. Normalizamos los datos pero con los **hiperparametros en memoria** y
luego con este nuevo $X_n$ ya tenemos $\theta_n$

$$X_n \theta_n = y' $$
Este $y'$ es el precio de la casa con 2000  y 5 cuartos.

### Algunas tecnicas de normalizaci'on

* **Estandarizacion** (standarization): Se **normaliza** los datos para que tengan media 0 y desviaci'on estandard $\sigma$. Como hacemos esto?

Definamos la media y desviacion estandard.
Cada columna $j$ tiene una media una desviaci'on estandard.

$$ \mu_j = \frac{\sum_{i=1}^m  x_{ij}}{m} $$
$$ \sigma_j = \sqrt{ \frac{\sum_{i=1}^m (x_{ij} - \mu_j)^2}{m}}$$

Como se modifica la matriz de datos $X$ para llegar a la matriz $X_n$

A cada columna $X_j$ de $X$
$$X_j :=  \frac{X_j - \mu_j}{\sigma_j} $$

La nueva matriz $X_n$ tiene media $0$ y desviaci'on estandard 1.

Este metodo es el mas comun.


* **min-max normalization** (normalizacion minimo maximo

$$X_j := \frac{X_j - \min_i\{x_{ij}\}}{\max_i \{ x_{ij} \} - \min_i \{ x_{ij}  \} }  $$
Esta tiene la ventaja de que acomoda todos los numeros de la columna en el intervalo $[0,1]$

* **mean normalization** (normalizacion por la media)
$$X_j := \frac{X_j - \mu_j}{\max_i  \{ x_{ij} \} - \min_i \{x_{ij} \}} $$

* **normalizacion vectorial**: Se divide por la norma del vector (por ejemplo la norma 2)

$$X_j := \frac{X_j}{ \| X_j \|}. $$

## Regresion polinomial tambien es regresion lineal

Asumamos para comenar que tenemos un curva polinomica

$$h(\theta, x) = \theta_0 + \theta_1 x  + \theta_2 x^2 + \cdots + \theta_n x^n $$
La curva sigue siendo lineal en $\theta$ (exponentes 1)
Podemos pensar que esto se escribe

$$h(\theta, x) = X \theta = \langle X , \theta \rangle  $$
donde
$$X=(1, x, \cdots, x^n) $$
y $\theta=(\theta_0, \theta_1, \cdots, \theta_n)^T$

Si se discretiza (se usa con datos) obtnemos
matricialmente

$$h(\theta, x) = X \theta  $$
mas exactmente

$$
\begin{pmatrix}
1 & x_1 & x_1^2 & \cdots & \cdots & \cdots & x_1^n \\
1 & x_2 & x_2^2 & \cdots & \cdots & \cdots & x_2^n \\
\vdots & \vdots & \vdots & \ddots & \ddots & \vdots \\
\vdots & \vdots & \vdots & \ddots & \ddots & \vdots \\
1 & x_m & x_m^2 & \cdots & \cdots & \cdots & x_m^n \\
\end{pmatrix}
\begin{pmatrix}
\theta_0 \\ \theta_1 \\  \vdots \\ \vdots \\ \theta_n
\end{pmatrix}
=
\begin{pmatrix}
y_1 \\ y_2 \\  \vdots \\ \vdots \\ y_m
\end{pmatrix}
$$

la matriz

$$V = \begin{pmatrix}
1 & x_1 & x_1^2 & \cdots & \cdots & \cdots & x_1^n \\
1 & x_2 & x_2^2 & \cdots & \cdots & \cdots & x_2^n \\
\vdots & \vdots & \vdots & \ddots & \ddots & \vdots \\
\vdots & \vdots & \vdots & \ddots & \ddots & \vdots \\
1 & x_m & x_m^2 & \cdots & \cdots & \cdots & x_m^n \\
\end{pmatrix} $$
se llama matriz de
[Vandermonde](https://en.wikipedia.org/wiki/Vandermonde_matrix).
Si $m=n+1$ la matriz es cuadrada y el problema tiene solucion

$$\theta = V^{-1} y $$
Esta soluci'on coincide con la interpolaci'on polin'omica de Lagrange.

Algunos hechos importantes con respecto al sobre/sub ajuste.

* En la practica tenemos muchos mas puntos $m$ que atributos $n$, es decir $m \gg n$. En este sentido es mas regresion que interpolation. La interpolaci'on puede inferir un costo mas alto con resultados que no **generalizan**.

* Si tenemos un numero de atributos muy grande (polinomio de orden 1000, o una red neuronal con 1000 capas) estamos tendiendo a hacer **overfitting**. Recuerden dos caracteristicas de los polinomios:
    * Oscilaciones de alta frecuencia (para grados altos)
    * Amplitudes grandes (sobre todo al final. Runge).

* El sobre-ajuste puede hacer que aprendamos mucho del ruido y luego al tratar de "generalizar" encontramos soluciones catastroficas.
* No queremos perder tiempo en modelos que no van a funcionar y son mas costosos.

### Como se puede controlar el "overfitting"
* Reducir el numero de atributos (features)
* Aplicar regularizacion. Esto lo vemos en la siguiente seccion.
Entre mas grande $\lambda$ mas suave es la solucion.
* Jugar con los datos y tener algunos guardados para evaluar el modelo.
    * Por ejemplo 70% entrenamiento 30% prueba
    * 60% entrenamiento, 20% cross-validation (validacion cruzada), 20% prueba.
    * $k$-fold validation (al final de clase ).

# $k$-fold cross-validation
Existen varios metodos para validar un modelo.

Sobre este tema volvemos mas adelante.
1. Partir los datos en dos grupos: 70%, 30%. El primero de entrenamiento y el segundo de prueba.
2. Partir en 3 grupos 60% , 20%, 20%: Entrenamiento, x-validation y prueba.

3. $k$-fold cross-validation.
La idea es:

Esto es para modelos con pocos datos.
1. Parta el modelo en $k$ grupos.
2. Entrena el modelo con $(k-1)$, se prueba con el modelo $k$ no escogido previamente.
3. Calcula el error en la generalizacion.
4. Se rota. Se escoge otro $i$ no escogido antes y se hace lo mismo.

5. Al final se tienen $k$ estimados de precision en los $k$ experimentos. Se promedia esta precision y esto da una idea de lo bien o no que se comporta el modelo.
