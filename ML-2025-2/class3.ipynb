 $$A^k = P \Lambda^k P^{-1} $$

 $$A^1 = P \Lambda^1 P^{-1} $$

 Asumamos
 $$A^k = P \Lambda^k P^{-1} $$

 \begin{eqnarray}

Se cayo la ficha $k+1$.

## Singular Value Decomposition (SVD): Descomposicion en valores singulares.
Dada $A \in \mathbb{C}^{m \times n} $, entonces existe una matriz ortogonal
$U \in \mathbb{C}^{m \times m}$, otra matriz ortogonal $V \in \mathbb{C}^{n \times n}$ y una matriz "diagonal" $\Sigma$ de orden $m \times n$, con valores no negativos tal que

$$A = U \Sigma V^*. $$

**Prueba** (de Cornelius Lanczos).
Asuamamos $A \in \mathbb{C}^{m \times n} $
Construimos una matriz en bloques.

$$S = \begin{pmatrix}
0  & | & A  \\
-- & | & -- \\
A^* & | & 0
\end{pmatrix} $$
Esta matriz $S \in \mathbb{C}^{(m+n) \times (m+n)}$
es **autoadjunta**.
tenemos $m+n$ autovectores $w_i$ y $m+n$ autovalores reales $\lambda_i$.

$$S w_i = \lambda_i w_i $$
Partimos $w_i$ de la siguiente forma
$$w_i = \binom{u_i}{v_i} $$
donde $u_i \in \mathbb{C}^m$ y $v_i \in \mathbb{C}^n$. Entonces tenemos que

Tenemos
\begin{eqnarray}
A^* u_i = \lambda_i v_i \tag{4} \\
A v_i = \lambda_i u_i \tag{5} \\
\end{eqnarray}

Multiplicamos  (4) por $A$ a izquierda y la segunda por $A^*$ a izquierda


\begin{eqnarray}
A A^* u_i = \lambda_i A v_i = \lambda_i \lambda_i u_i = \lambda_i^2 u_i  \\
A^* A v_i = \lambda_i A^* u_i= \lambda_i \lambda_i v_i = \lambda_i^2 v_i  \\
\end{eqnarray}

Es decir que $\lambda_i^2$ son autovalores de $A^*A$  y $A A^*$.
Los autovectores de $A A^*$ son los $u_i$
Los autovectores de $A^* A$ son los $v_i$

Es posible que algunos  autovalores sean $0$ y que queden $r$ que no son cero. Los organizamos $\lambda_1 \ge \lambda_2 \ge \lambda_r>0$. $\lambda_i=0$, $i>r$
Llamamos a la matriz $U_r$ aquella formada por $r$ columnas de los $u_i$
Llamamos a la matriz $V_r$ aquella formada por $r$ columnas de los $v_i$

La matriz $U_0$ cuyas columnas corrseponden a autovalores $0$.
La matriz $V_0$ cuyas columnas corrseponden a autovalores $0$.
$$ U = [ U_r | U_0]$$
$$ V = [ V_r | V_0]$$

De las Ecuaciones (4) y (5)

\begin{eqnarray}
A^* U_r &=& V_r \Lambda_r  \quad , \quad A^* U_0 = 0 \\
A V_r &=& U_r \Lambda_r \quad, \quad  A V_0 = 0.
\end{eqnarray}

$\Lambda$ es diagonal con los $\lambda_i^2$, $\Lambda_r$ son los que no son 0,
$\Lambda_0$ los que son 0.

Veamos la siguiente operacion por bloques

$$A V = A [ V_r | V_0] = [ A V_r | A V_0 ] =[ U_r \Lambda_r | 0 ] = [ U_r \Lambda_r + U_0 \times 0 | U_r \times 0 + U_0 \times 0 ] = U \Sigma $$

Donde
$$\Sigma = \begin{pmatrix} \Lambda_r &|& 0 \\ -- & -- & -- \\ 0 &|&  0
\end{pmatrix} $$

De aca que
$$AV = U \Sigma $$
Podemos multiplicar a derecha por $V^*$

$$AVV^* = U \Sigma V^*$$
poro $V V^* = i$

$$A = U \Sigma V^*$$

$\lambda_i^2$ son autovalroes de la matriz $S$ que es simetrica, positovos 0 0.

$$\sigma_i = \sqrt{ \lambda_i^2} = | \lambda_i |  $$
A estos se les llama **valores singulares**.

Ejemplo:
Encontrar la descomposicion SVD de la matriz

\begin{pmatrix}
1 & 0 & 3 \\ 1 & 2 & 4
\end{pmatrix}

Muestro el algoritmo
* Construya $AA^*$, $A^* A$
* Halle los autovectores de $A A^*$ y autovalores
 Halle los autovectores de $A^* A$ y autovalores
* Los autovectores de $A A^*$ forman la matriz $U$
 Los autovectores de $A^* A$ forman la matriz $V$

* Algunos autovalores podrian ser 0. LLame $\sigma_i= \sqrt{\lambda_i^2}$.
en palabras, halle los autovalores de $A^* A$ y les saca la raiz cuadrada y esos son los $\sigma_i$.
Ordene $\sigma_1 \ge \sigma_2 \ge \cdots \ge \sigma_r > 0 $.
$\sigma_i = 0$, $i > r$.

El orden en que se coloquen los $u_i$ y los $v_i$ es importante. Piense acerca de esto.




## Cociente de Rayleigh.

**Definicion 2.3.2** (Cociente de Rayleigh). Asuma que la matriz $A$ es autoadjunta. Definimos el **cociente de Rayleigh** de $A$ con respecto al vector $x \in \mathbb{C}^n$ como

$$R(A, x) = \frac{x^* A x}{x^* x} = \frac{\langle x, Ax \rangle }{\| x \|^2} = \frac{ \langle Ax, x \rangle}{\| x \|^2}  $$

Tambien dado que $u=x/\| x \|$,

$$\frac{\langle x, A x \rangle }{\| x \|^2} = \langle u, A u \rangle $$

Para hallar el cociente de Rayleigh, elegimos $u$ en una esfera unitaria $\| u \|=1$ y probamos $\langle Au, u \rangle $



```python
import numpy as np
import matplotlib.pyplot as plt

# generate points in a unit circle
theta = np.arange(0, 2*np.pi, np.pi/24)
circle = np.array([np.cos(theta), np.sin(theta)])

# definamos la transformacion
A = np.array([[1, 0.5],[0.5,1]])

# aplicamos el Mapping
transformed = A @ circle

# origen
origin_x = np.zeros(circle.shape[1])
origin_y = np.zeros(circle.shape[1])

# grafica
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect('equal')

# draw original vectors in black
ax.quiver(origin_x, origin_y, circle[0], circle[1], color='black', label='original vectors',
          scale_units='xy', angles='xy', scale=1, alpha=0.6)

# draw mapped vectors
ax.quiver(origin_x, origin_y, transformed[0], transformed[1], color='red', label='mapped vectors',
          scale_units='xy', angles='xy', scale=1, alpha=0.6)

# anotar autovectores
ax.text(1.4, 0.0, '$\\vec{u}$', fontsize = 12)
ax.text(0.0, 1.1, '$\\vec{v}$', fontsize = 12)

# Axis
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

ax.grid(True)
ax.title("autovectores...")

plt.tight_layout()
plt.show()


```

```python
def Rayleigh(A, theta):
    return np.array ( [ np.array([np.cos(t), np.sin(t)])  @ A @ np.array([np.cos(t), np.sin(t)]) for t in theta])

```

```python
# plot the Rayleigh quotient
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect('equal')

plt.plot(theta, Rayleigh(A, theta), color='red', label='Rayleigh quotient')
plt.show()
```

**Teorema 2.3.2**(Rayleigh Quotient). El mas grande (largest) de los autovalores de una matriz adjunta $A$ es el mayor de los cocientes de Rayleigh, tomados sobre una esferan unitaria.

**Prueba**:
Asuma $u \in \mathbb{C}^n$ es un vector unitario. Del teorema de descomposicion espectral se puede escribir $u$ en terminos de una base ortogonal que son los autovectores de $A$

$$u = \sum_{i=1}^n \alpha_i u_i  $$
donde $u_i$ es el $i$ autovector unitario de $A$.

De aca que

$$R(A, u) = u^* A u = \sum_{i=1}^n \sum_{j=1}^n \overline{\alpha_i} \alpha_j u_i^* A u_j  \tag{5} $$

Ahora bien
$$u_i^* A u_j = u_i^* (\lambda_j u_j) = \lambda_j u_i^* u_j = \lambda_j \delta_{ij}  \tag{6} $$
Insertamos 6 en la 5

$$R(A, u) = \sum_{i=1}^n \sum_{j=1}^n \overline{\alpha}_i \alpha_j u_i^* A u_j = \sum_{i=1}^n \sum_{j=1}^n \overline{\alpha}_i \alpha_j \lambda_j \delta_{ij} =  \sum_{i=1}^n | \alpha_i |^2 \lambda_i $$

Ahora como $\sum | \alpha_i |^2 = 1$
Sin perdida de generalidad puedo escoger $\lambda_1$ como el mayor de todos los autovalores. Como $R(A, u)$ es un promedio ponderado de los $\lambda_i$ con pesos $| \alpha_i |^2$, y estos pesos suman 1. Si $|\alpha_1|=1$ y
$|\alpha_i|=0$, $i=2,3, \cdots , n$ estoy asegurando el mayor de los posibles valores de $R(A, u)= \lambda$.
