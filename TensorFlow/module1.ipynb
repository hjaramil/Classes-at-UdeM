# Rank/Degree of Tensors

```python
rank1_tensor = tf.Variable(["Test"], tf.string)
rank2_tensor = tf.Variable([["test", "ok"], ["test", "yes"]], tf.string)
```

```python
tf.rank(string)  # rank 0, same as int and float above
print("\n")
tf.rank(rank1_tensor)  # rank 1
print("\n")
tf.rank(rank2_tensor) # rank 2
```

## Shape/reshape/flatten

```python
rank2_tensor.shape  # it is a 2x2 matrix, same as numpy
print("\n")
rank1_tensor.shape
```

```python
tensor1 = tf.ones([1,2,3])
tensor1
```

```python
tensor1.shape
```

```python
tf.rank(tensor1)
```

```python
tensor2 = tf.reshape( tensor1, [2,3,1])
tensor2
```

```python
tf.rank(tensor2)
print("\n")
tensor2.shape
```

```python
tensor3 = tf.reshape(tensor2, [3,-1])  #-1 tells the tensor to calculate the size of the
                                       # dimension in place. That is reshape into [3,2]
tensor3
```
