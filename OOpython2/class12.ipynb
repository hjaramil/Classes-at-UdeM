### Un ejemplo simple

## `try-else`  block
El objetivo es imprimir "bye" con o sin excepcion

```python
def divideMe(a,b):
    try:  # we try to divide
        c=a/b  # critical statement
        print(c)
        return
    except Exception:
        print("you are trhing to divide a number by 0 and this is not allowed")
        return

a=5
b=0
divideMe(a,b)
print("bye")
```

```python
b=3
divideMe(a,b)
print("bye")
```

Para que necesitamos nosotros que el flujo del programa siga bajo una excepcion critica. Muchas veces nosotros separamos recursos del sitema,
por memoria, o abrimos un archivo, u otros. Por ejemplo, si abrimos archivo lo debemos cerrar. Navin Redy, dice "cuando te levantas a media noche y abres la nevera para sacar una salchicha, la vuelves a cerrar". Asi debe ser con los archivos y con los recursos que se separen del sistema (memoria u otros).
Si se produce una excepcion (error fatal) cuando un archivo se abrio, ese archivo queda abierto y eso es bueno. 

### mostremos tambien cuales errores que arrojan las excepciones



```python
def divideMe(a,b):
    try: 
        print("abrimos un archivo") # un simulacro
        c = a/b
        # hacemos un proceso
        print("cerramos el archivo")
    except Exception as er:  # esto captura en la variable "er" el erro que el sistema genera
        print(er)
    return

a=5
b=3
divideMe(a,b)
print("bye")
```

```python
b=0
divideMe(a,b)
print("bye")
```

## `try-except-else` block




```python
# el else se ejecuta solo si no hay excepciones
def divideMe(a,b):
    try: 
        print("abrimos un archivo")
        c=a/b  # critical
    except Exception as er:  # "er" es el error del sistema
        print(er)
        print("cerramos el archivo")
    else: # cuando no hay exepcion
        print("cerramos el archivo luego de hacer un trabajo")
    return

a=5
b=0
divideMe(a,b)
print("bye")
```

```python
b=3
divideMe(a,b)
print("bye")
```

## El `try-except-else-finally`  block


```python
import queue
def divideMe(a,b):
    try:
        print("abrimos un archivo")
        c=a/b
    except Exception as er:
        print(er)
    else:
        print("cerramos el archivo luego de un trabajo")
    finally:  # Este se hace, o se hace, no importa que
        print("esto se hace no importa que")
        print("por ejemplo puedo cerrar el archivo aca")
    return

a=5
b=0
divideMe(a,b)
print("bye")
```

```python
a=5
b=2
divideMe(a,b)
print("bye")
```

Aca ocurrio algo que no debe ser. Cerramos el mismo archivo dos veces.
Tambien puede ocurrir que vaya a abrir un archivo que ya este abierto, o que no exista.




```python
# usamos la idea de las "banderas" (flags)
def divideMe(a,b):
    flagOpen = 0
    try:
        if not flagOpen:
            print("abramos el archivo")
            flagOpen  =1 
        c=a/b
    except Exception as er:
        print(er)
    else:
        if flagOpen:
            print("cerramos el archivo que estaba abierto")
            flagOpen=0
    finally:
        print("esto se hace, o se hace")
        if flagOpen:
            print("cerramos el archivo que estaba abierto")

    return

a=5
b=1
divideMe(a,b)
print("bye")
```

```python
b=0
divideMe(a,b)
print("bye")
```

## Assertions:  (corroboraciones)
Son como excepciones pero se usan para hacer "debuggin". La sintaxis es la
siguiente:

`assert Expression[, Arguments]`

Veamos algunos puntos sobre `Assert`.

* `assert`  retorna un valor booleano. (se espera que sea cierto).
* `assert`  tiene un mensaje opcional que usted puede formular para ayudar en el proceso de "debugging"
* `assert` se usa para verificar tipos de salidas (enteras, booleanas, otros), unidades, signos, etc.

Ejemplos



```python
def avg(speeds):
    n = len(speeds)
    assert n !=0 
    return sum(speeds)/n

vels = []
print("promedio de velocidades", avg(vels))
```

```python
def avg(speeds):
    n = len(speeds)
    assert n !=0, "Verifique que su arreglo de velocidades no es vacio"
    return sum(speeds)/n

vels1 = [1,2,3,5]
vels2 = []
print("promedio de velocidades", avg(vels1))
print("promedio de velocidades", avg(vels2))
```

```python
class Car:    

    def __init__(self, model, year):
        self.model = model
        self.year = year
        return

    def __str__(self):
        return f'from str: My car is a {self.model} of the year {self.year}'

    def __repr__(self):
        return f'from repr: My car is a {self.model} of the year {self.year}'

c = Car("Toyota", 2012)
print(c)
```

```python
class Car:    

    def __init__(self, model, year):
        self.model = model
        self.year = year
        return

    

    def __repr__(self):
        return f'from repr: My car is a {self.model} of the year {self.year}'

    def __str__(self):
        return f'from str: My car is a {self.model} of the year {self.year}'

c = Car("Toyota", 2012)
print(c)
```
