Regresando al OO (Object Oriented). Por que esta tecnologia es importante.

* Es **natural**. En el sentido en que en la vida nosotros somos y estamos rodeados de objetos y pensar en objetos es natural.

* Es **confiable** (reliable). Como el lenguaje esta en modulos (clases) se pueden arreglar unas cosas sin danhar otras (aunque los modulos estan conectados y hay que tener cuidado con esto).

* Se puede reusar. Es **reciclable** (reusability). Si por que una clase es un mapa para cualquier cantidad de objetos que quieras modelar. Y aca no para la cosa. Existen tecnicas (que las vamos a  ver) como herencia y polimorfisco que tambien indican como reusar codigo. Esto simplifica mucho los codigos y los procesos.
Esto implica ahorro de tiempo, menos codigo, mas claridad, mas eficiencia , ahorro en memoria, ejecucion, etc.

* Es **extensible**: Simplemente puedes agregar clases sin que quiebre nada. 

* Es **flexible**. Por ejemplo yo puedo trabajar en una parte del codigo (una clase) mientras que un companero esta trabajando (en China) en otra clase sin que esto afecte "mucho" el funcionamiento global. Por que "mucho" entre comillas? Existe la interface (que no debe cambiar) y la implementacion donde estan los detalles y hay que tener cuidado con la inteface por que es la que nos comunica con el mundo exterior.



## Objetos y Clases:
### Objetos: 
Los objetos en la ciencia de la computacion no son muy distintos a los que vemos en la calle. Excepto que los objetos en la computacion son abstractos. Un carro en mi programa de computo es un modelo, una descripcion (detallada o no) del carro que va por la calle (el fisico, concreto). 

Un objeto es parte de un sistema y por lo mismo el sistema es una coleccion de objetos (y sus relaciones). En la descripcion de un objeto estan los atributos y los comportamientos. Los **atributos** son los sustantivos y adjetivos que describen el objeto. Los comportamientos son verbos, el carro "move()". Se desplaza, el perro "corre()". 
En resumen: los objetos tienen **atributos** y **comportamientos**. Los comportamientos se modelan con **funciones** que en la literatura de OO se llaman **metodos**.

Los objetos se comunican entre si: Se comunican a traves de mensajes. Para pasar un mensaje (funciones) se necesita el nombre de mensaje y los parametros. 

### Clases.
Agreamos una lista de propiedades que deben tener las clases:

* Una clase es el mapa (blueprint) o plano de un objeto. La clase es una "abstaccion" del objeto. Es una patente del objeto. 

* Es un tipo (type).

### Polimorfismo:
El polimorfismo consiste en varias formas. El polimorfismo nos permite reciclar codigo. Algunos objetos o atributos, o comportamientos, etc tiene parecidos y se pueden todos conjugar bajo el mismo nombre con distintas acciones.

* Polimorfismo a traves de herencia.
Piense en la clase ```Animal``` . Alli podemos definir un metodo llamado ```mover()``` . Pero hay muchos tipos de animales. Peces, humanos, aves, reptiles.Cada uno se mueve de distinta forma. Podemos usa parametros como A, B, para moverse desde A hasta B. Pero en la implentacion de cada animal esta la forma de moverse (nadar, caminar, volar, arrastrase). Este tipo de polimorfismo se llama "runtime polimorphism" o dinamico.

* Polimorphismo estatico: Por ejmplo "method overloading" (sobrecarga de metodo). Por ejemplo sum(a,b,c) o sum(a,b). De acuerdo al numero de parametros se ejecuta a+b+c  o a+b 
Operator overloding:  A+B puede ser la suma de dos escalares o dos matrices o dos numeros complejos.
