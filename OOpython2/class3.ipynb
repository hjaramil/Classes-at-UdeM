Vamos a motivar el uso de OOP (Object Oriented Programming).

Algunas ventajas:

* Modularity (modularidad). El OOP se divide en modulos lo que lo hace mas facil de entender y manejar. "Divide an Conquer". ("une y gobernaras"). Para meantenimiento es mejor. 

* Simplificacion de codigo. 
[Bjarne Stroutstrap](https://www.youtube.com/watch?v=uTxRF5ag27A) (el inventor de C++) dice. Que en OO la complejidad se incrementa con el numero de lineas de codigo, mientras en procedural programming (programacion en procedimientos) la complejidad aumenta con el cuadrado de numero de lineas en el codigo. Esto no es un teorema matematico. Es una forma de hablar. 

Explico en la pizarra......

* Reusability (reusabilidad). El codigo se puede reciclar. O SEA, por ejemplo, usted define una clase, y esa clase la puede usar para generar cuantos objetos quiera, sin necesitad de re-escribir la clase. Otras formas de reciclar son la herencia y el polimorfismo, donde se puede reciclar codigo elegantemente.

* Seguridad: En ingles existen dos palabras: security, safety. La seguridad es que tan protegido este un sistema contra "hacking", violacion de candados, etc. Que tan seguros son sus datos. Pero "safety" es el riesgo de resultar herido, o que el programa se "corrompa". Las dos cosas "safety" , "security" son atributos faciles de implementar en OOP.

## Historia:
DAtos en bullets:

* El lenguaje **SIMULA** fue el primero (de los 1960s) que introdujo los conceptos de clase, hererencia, y otros atributos en OOP. Se baso en ALGOL 60. SUMULA fue inventado en un centro de computo en OSLO (Noruega) por Ole-Johan Dahl y Kristen Nygaard.

* Bjarne Stroustrup (el inventor de C++). Hace el siguiente comentario. "Un sistema de trafico en Noruega), piense en una senal de transito, que si es carro gire a la izquierda, que si es una bicicleta, gire a la derecha, que si es un carro de bomberos....etc, etc."  Esto es muy ineficiente de esta forma en un programa de procedimientos (necesita una cantidad de "if" statments). 

* Alan Kay, trabajo en XEROX. Es importante por que inventaron los GUIs: Graphical User Interface. 
Steve Jobs de Apple creo su sistema Macintosh de los GUIs, lo mismo que Windows 95 (Bill Gates). 
Alan Kay fue el que acuñó la frase " Object Oriented".
Esto lo hizo en el contexto del lenguaje SmallTalk. 

* **Laboratorios Bell**: Es una empresa de mucho tiempo. Ha tenido mas 6 premios nobel. Se invento el lenguaje C, Dennis Ritchie. Con el C se hizo Unix, quien? Ken Thompson, tambien basado en "B" .  Despues  Bjarne Stroustrup, extendio el C a C con clases, pensando en SIMULA. Pero mas adelante no se quiso quedar con eso solamente he invento el C++.

* Como aparecio Java. El C++ manejaba la memoria con detalle, pero este manejo es complicado. El nuevo lenguaje Java uso una tecnica llamada "garbage collection" (recoleccion de basura) en la cual usted como programador no tenia que preocuparse por liberar la memoria "allocated" (separada, guardada). La sintaxis de Java es muy parecida a la de C++ . Este lenguaje fue inventado por James Goslin. 
Y Java se volvio el lenguaje mas popular por casi 20 anhos.

Este es el lenguale del NeXT. Objective C fue el lenguaje de Apple uso para desarrollar su sistema operacional MacOS.
