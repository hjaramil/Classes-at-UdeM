Entramos a estudiar **exceptions**.

Las exceptciones son mensajes que el sistema le manda al usuario acerca de algo inusual. Algun error, una advertencia. Lo para bien. La excepciones son una ayuda para diagnosticar problemas en su programa. Piense en el dolor en un ser humano. Hay muchas formas de clasificar las excepciones. De acuerdo al estado del programa

* **fatales** : fatal error. El programa para, ejemplo el programa anterior.
* los fatales arrojan una advertencia y siguen. Ejemplo, en el proyecto de geometria lanzar una excepcion si 3 puntos estan alineados y decir que el poligono es degenerado.

En general, pueden hacer 3 tipos de excepciones:

1. Hardware failure. Un problema en la memoria de un chip.
2. Un error no fatal, que manda un mensaje y sigue corriendo
3. Software excepcion. Los ejemplos de arriba. Division por 0,
raiz de un numero negativo y muchas mas. Algunas de ellas vienen
con el sistema (built-in) y otras son definidas por el usuario.

Cuando una excepcion se genera se dice (en ingles) que sistema es
**throwing an exception**.

### User defined (custom) exceptions.
Cuando desarrollamos software debemos tambien avisarle al usuario de las "exceptions". Por ejemplo trabajando con masas, si el usuario ingresa una masa negativa, reportarle que esas no existen.

La excepciones se pueden definir en un archivo separado, por ejemplo con nombre `myExceptions.py` . 

Veamos ejemplos.

```python
class AgeLegalToDrinkError(Exception): 
    """ Exception is raised from errors in the age in a bar.

    Attributes:
    age -- input age which caused the error
    message -- explnation for the error
    """ 


    def __init__(self, age, message='age needs to be  over 21 years old'):
        self.age = age
        self.message = message
        return

    def __str__(self):
        return f'{self.age} -> {self.message} '

age = int(input("Enter age: "))
if not age > 20:
    raise AgeLegalToDrinkError(age)
    
```
