### Un ejemplo con ```__new__()``` y ```__init__()``` .
(dunders: double underscore).

````python
class Car:

    def __new__(Clase, brand, model, year):
        print(f'Creating a new {Clase.__name__} object...')
        obj = object.__new__(Clase)
        print(id(obj))
        return obj


    def __init__(self, brand, model, year):
        print(f'Initializing the Car object ...')
        self.brand  = brand
        self.model = model
        self.year = year
        return

myCar = Car('Toyota', 'Colorla', 2010)
print(id(myCar))
print(myCar)

````

## Encapsulacion
Recordemos un poco de la teoria. Meter toda la informacion en "capsulas". 
La informacion son atributos y metodos. Se tienen algunas ventajas de esto:

* Ocultar informacion: Se puede ocultar informacion de objetos externos y aun de objetos de la misma clase. Lo vemos en Python con ejemplos.
Esto tiene varias ventajas:

    *  Protejer sus datos: Nosotros mismos nos podemos hacer dano. nos protejemos de que otros vean nuestra informacion confidencial. Por negocio o por "verguenza". Propiedad intelectual.
    * Protejer a los otros. En una carro es importante que las direccionales funcionen y esten activadas cuando vamos a girar. Pero a los conductores externos no les importa la temperatura de su motor, y las RPM. Imaginese todos los condutores viendo pantallas imaginarias con todos los datos del funcionamiento del motor de cada carro. Seria un distractor y representaria peligro. 

* Modularidad: Los modulos simplifican los programas. 

* Reuso (reciclaje). Los objetos se deben definir una vez con sus atributos sin necesitad de re-escribir codigo. Se intancian cuantas veces sea necesario con el mismo codigo. Si hay un cambio de textura entre objetos se puede usar herencia y/o polimorfismo. Veremos ejemplos

* Ocultar complejidad (recuerden el telfeno transparente).

Los lenguajes como Java , C++ ocultan atributos con variables privadas,
Python dice si tu variable comimenza con un guion bajo (_myVariable, por ejemplo) es privada. Sin embargo Python te la muestra si la quieres ver. O sea que esto es una convencion para que entre nosotros no nos "pisemos las mangeras". Si queremos ocultar una variable le ponemos doble guion bajo al comienzo. Vamos a ver que a esto tambien se le puede dar la vuelta y a esto se la llama *name mangling".

Los lenguajes Java, C++ permiten ver los datos privados a traves de intefaces tales como "getters" y "setters" (los getters son para ver, los setters para cambiar (estos ultmos se llaman "mutator").

Vamos ejemplos de "getters", "setters" pero vamos a ver tambien que Python no los necesita. 



````python
class Car:

    def __init__(self, brand, model, year, engineTemperature):
        self.brand  = brand
        self.model = model
        self._year = year   # se supone privado. Stay out!
        self.__engineTemperature = engineTemperature  # se supone oculta
        return

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1

        return

myCar = Car('Toyota', 'Colorla', 2010, 20)
myCar.tempetaturePlusOne()

print(myCar.brand)
print(myCar.model)
print(myCar._year)
````

## como podemos ver aquellos atributos "ocultos" ?
Inicialmente usamos los getters y setters de Java y luego vemos otras formas.


````python
class Car:

    def __init__(self, brand, model, year, engineTemperature):
        self.brand  = brand
        self.model = model
        self._year = year   # se supone privado. Stay out!
        self.__engineTemperature = engineTemperature  # se supone oculta
        return

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1

    # getters
    def getTemp(self):
        return self.__engineTemperature

    # setter
    def setTemp(self, newtemp):
        self.__engineTemperature = newtemp
        return

myCar = Car('Toyota', 'Colorla', 2010, 20)
myCar.temperaturePlusOne()

print(myCar.brand)
print(myCar.model)
print(myCar._year)
````

## Argumentos por defecto (default arguments)


````python
class Car:

    def __init__(self, brand="Toyota", model="Corola", year=2021, engineTemperature=100):
        self.brand  = brand
        self.model = model
        self._year = year   # se supone privado. Stay out!
        self.__engineTemperature = engineTemperature  # se supone oculta
        return

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1

    # getters
    def getTemp(self):
        return self.__engineTemperature

    # setter
    def setTemp(self, newtemp):
        self.__engineTemperature = newtemp
        return

myCar = Car()
myCar.temperaturePlusOne()

print(myCar.brand)
print(myCar.model)
print(myCar._year)
````
