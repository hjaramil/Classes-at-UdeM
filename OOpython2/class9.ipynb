sobreescritura es parte de un mecanismo de OOP llamado **polimorfismo**.
Pero,  sin embargo, overrding es tambien parte de **inheritance**.
"overriding" esta en la interseccion entre polimorfismo y herencias.

Lo vemos en este topico, pero tambien lo veremos dentro del tema de polimorfismo. 

Que es overrding? Es un metodo que se hace en la clase derivada conservando el nombre pero cambiando su funcion. 

Ejemplo de overriding (overwriting)

```python
# en colombia la edad legal para beber alcohol en un negocio es 18 
# anhos, en USA es 21. Entonces hacemos un metodo
# que en la clase madre es como si fuera colombia 
# y en la clase hija como si fuera USA.

class Person:
    def __init__(self, name , age, inBar):
        self.name = name
        self.age = age
        self.inBar = inBar
        return

    def adult(self):
        if self.age >= 18: # en colombia
            return True
        return False  # sobra el else

    def fineBarOwner(self):  
        if not self.adult()  and self.inBar:
            print("%s, you should not be here"%self.name)
            print("You and the bar owner got problems")
            return
        print("%s, you are fine. You are a good guy"%self.name)


    
class Drinker(Person):
    def __init__(self, name, age, inBar):
        super().__init__(name, age, inBar)
        return

    def adult(self):  # overriding/overwriting
        if self.age >= 21:  # en USA
            return True
        return False  # no se requiere ELSE aca

    def fineBarOwner(self): 
        super().fineBarOwner()
        return


jesus = Person("Jesus", 15, True)
jesus.__dict__





```

```python
jesus.fineBarOwner()
```

```python
rob = Person("Roberto", 18, True)
rob.__dict__
rob.fineBarOwner()

```

```python
robert = Drinker("Roberto", 18, True)
robert.__dict__
#robert.fineBarOwner()
```

```python
robert.fineBarOwner()
```

## Multiple inheritance (herencia multiple)
Se hereda de dos clases distintas


```python
class Merchandise:

    def __init__(self, number, cost):
        self.number = number
        self.cost = cost
        return

    def computeTotalCost(self):
        total = self.number*self.cost
        return total

class Fruit:

    def __init__(self , name):
        self.name = name
        return


class Item(Merchandise, Fruit):  # herencia multiple
    def __init__(self, name, number, cost):  # init gets all arguments from both parents
        Fruit.__init__(self, name)
        Merchandise.__init__(self, number, cost)
        return

orange = Item("Orange", 300, 23.5)
orange.__dict__  ## es lo mismo que vars(orange)
```

```python
orange.computeTotalCost()
```
