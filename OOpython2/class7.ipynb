````python
class Car:
    brand="Toyota"
    mode="Cololla"
    year=2014

    def __init__(self, color="yellow"):
        self.color = color
        return

    def stop(self):
        print("por favor pare")
        return

````

## Interaccion entre objetos
### Asociacion:
* Composicion
* Agregacion

````python
# ejemplo de composicion:
class Point:
    def __init__(self, x, y ):
        self.x  = x
        self.y = y
        return

    def norm(self):
        return ( self.x**2 + self.y**2)**0.5
````

````python
# clase segmento
# por ahora 4 coordenadas, mas tarde 2 puntos
# es solo por ilustrar "composicion"
class Segment:
    def __init__(self, x1,y1, x, y ):
        self.x1 = x1
        self.y1 = y1 

        # fijamos el segmento en el origen
        self.x = x - x1
        self.y = y - y1

        # hagamos composicion
        self.point = Point(self.x, self.y)
        return

    def lengthSegment(self):
        return self.point.norm() 

# probemos
seg = Segment(3,4,5,6)

# hallemos la longitud del segumento
print(seg.lengthSegment())

````

````python
# probemoslo:
import numpy as np
p1 = np.array([3,4])
p2 = np.array([5,6])
dist = np.linalg.norm(p1-p2)
dist
````
