# Scope: Alcance de las variables.
Alcance es tiempo de vida de una varaible en el programa.

* Local
La variable esta definida dentro una funcion
* Global
Teine el alcande del archivo
* built-in
Vive mientras la sesion de Python este vigente.

Ejemplos:


### Local Scope

````python
# La varible pi es local en este codigo
def areaDelCirculo(r):
    a=5*r
    pi=3.14159 # la variable pi nace aca
    print(2*a)
    return r**2*pi
    # la variable a muere aca

print(areaDelCirculo(2))
# namespace
print(a)
````

### Global Scope



````python
def areaDelCirculo(r):
    global a  # dura mientras estemos en este archivo
    a=2*r
    pi=3.14159
    print(2*a)
    return pi*r**2  # la vida de pi sigue por fuera de esta funcion.

print(areaDelCirculo(2))
print(a)
print(pi)
````

Que diablos es ```self``` ?

El ```self`` es la direccion en memoria del objeto. 
Veamos por que:

````python
class Square:
    side = 2  # el lado, por defecto

    def sqr(self):
        print("la memoria de el objeto es", id(self))
        print()
        return self.side**2  # retorno el area

mySqr = Square()
mySqr.sqr()
print(id(mySqr))
````

La memoria del objeto es "self" y todo relativo a este objeto tiene como 
centro de coordenadas esa memoria "self". No hay nada especial con la palabra "self" pudimos hacer todo con otra palabra, por ejemplo "this:


````python
class Square:
    side = 2  # el lado, por defecto

    def sqr(this):
        print("la memoria de el objeto es", id(this))
        print()
        return this.side**2  # retorno el area

mySqr = Square()
mySqr.sqr()
print(id(mySqr))
````

Hay forma de inicializar objetos cuando se instancie la clase. Esto se hace
con el metodo ```__init__()``` (que es un constructor). 
Ejemplo.

````python
# queremos definir un cuadrado con atributos, lado, centro, color
class Square:

    def __init__(self, side, color, center):
        self.side=side
        self.color=color
        self.center=center
    def sqr(self):
        print(" La clase cuadrado se instancio con color=%s,  \
         \n lado=%f, y \n centro=%s"%(self.color, self.side, self.center))
        return self.side**2


mySqr = Square(2, "amarillo", (2,3))
print("el area del cuadrado es", mySqr.sqr())
````

A metodo ```__init__``` se le llama el constructor de Python. Realmente hay un constructor de python oculto que se llama el ```__new__()```

````python
# Como el metodo __new__ es llamado interamente para construir __init__
class Car:

    def __new__(coco, brand, model, year):
        print(f' Crear una clase nueva {coco.__name__} objeto')
        obj = object.__new__(coco)
        return obj

    def __init__(self, brand, model, year):
        print('Inicializandod el objeto Car...')
        self.brand = brand
        self.model = model
        self.year = year

myCar = Car("Toyota", "Corola", 2012)

````
