Para la logica existe un lenguaje llamado ProLog: Programcion logica.
Ejemplo: sume los elementos de una lista

``` 
sumlist([], 0).
sumlist( [H:T], N) :-   sumlist(T, N1), N is N1 + H.
```

Si lo correcmos con la lista ``` [1, 3, 5, 5] ``` obtenemos

```
sumlsit ( [ 1,3,5,4])
N=13
```

Depueremos este program en ProLog.
Datos de entrada ``` [1,3,5,4]```

* Inicialmente el conjunto (los datos de entrada) no es vacio. 
* Dividimos los datos en cabeza/cola ```H=1,  T=3,5,4]```
No sabemos inicialmente quien es ```N1```. Pero la instruccion dice
```N = N1 + 1 ``` (1)
La nueva lista es ``` [3,5,4]```

* Dividimos los datos en cabaeza/cola ```H=3, [5,4]```. La instruccion dice
``` N = N1 + 3 ``` (2)
* La lista es ahora ```[5,4]``` que no es vacia .  cabeza/coloa ```H=5, [4]```. De forma que ``` N= N1 + 5 ```  (3) 
* La lista se reduce a [4], la cabeza es ```H=4, T=[]```.
La operacion es ```N= N1 + 4```  (4).
* Finalmente en proceso hacia adelante (forward) llegamos que la lista es vacia,  De forma que ```N1=0```.  Ahora nos tenemos que devolver.

* De la ecuacion (4), y llamando ```N1``` a ```N```, ```N1 = 0+4 = 4```
* De la ecuacion (3), y llamando ```N1``` a ```N```, ```N1=4+5=9
* De la ecuacion (2), y llamando ```N1``` a ```N```, ```N1=9+3=12```
* De la ecuacion (1), y llamando ```N1``` a ```N```, ```N1=12+1=13

* functional programming (programacion funcional).
El paradigma funcional tiene ciertas caracteristicas.

* Esta basado en funcion puras. Son funciones en el sentido matematico, que usted entra unos parametros y tiene un solo valor para esos para estos parametros. 
* No tiene efectos colaterales. No se pueden cambiar los valores de las variables definidas. Esto es limitacion pero tambien ventaja, para debugging, para paralelizacion. La sincronizacion se hace mas*  facil.
* Se recomienda la recursion (yo diria que se obliga).
* Las variables son inmutables (esto repite lo que dije arriba).

Podemos hacer una lista de ventajas y desventajas.

* Ventajas:
    * Facil de depurar (debug).
    * La ejecucion podria ser eficiente mediante el uso de cache.
    Las funciones en el lenguaje funcional, como retornan un valor unico para el mismo conjunto de parametros, entonces pueden generar un k up table" (guardar en memoria de "cache" los resultados de los parametros ya usados y en vez de calcular nuevamente el valor, lo lee del "cache". 
    * Facil para revisar las entradas y las salidas.

* Desventajas.
    * Podrian resultar muchisimas copias de objetos muy similares.
    Ejemplo en la pizzarra Write. Piense en una lista que crece, los nombres de la lista deben cambiar para poder cumplir con la "inmutatibilidad". Pero esto genera mucha memoria.
    * Las recursiones son bonitas, elegantes pero pueden generar mucha memoria y hacer que el programa se mas lento. 

Hacemos ejemplos en Python. Python es "multiparadigma" que en Python se
puede hacer funcional, estructurado, procedimientos, OO.



````python
# en el imperativo las variables pueden ser mutables.
a =5
def sumeuno(x):
    return x+1

def sumeuno3veces(x):
    x = sumeuno(x)
    print("x=" , x)
    x = sumeuno(x)
    print("x=" , x)
    x = sumeuno(x)
    print("x=" , x)
    return x

print("a antes de ser cambiado",a)
a = sumeuno3veces(a)
print("a luego de cambiar", a)
````

#### Ejemplo anterior que se acoje al paradigma funcional

````python
def sumeunotresvecesfun(a):
    b = sumeuno(a)
    print("b=",b)
    b = sumeuno(a)
    print("b=",b)
    b = sumeuno(a)
    print("b=",b)
    return a

a=5
print("el valor de a es ",a)
a= sumeunotresvecesfun(a)
print("a ahora es",a)
````
