# Containers (contenedores).
Que es un contenedor. Es una estructura de datos que contiene varios objetos.
 Hay muchos tipos de contenedores. Los contenedores "built-ins". (add-ons).

 * **lists**: 
    * Es mutable (se puede cambiar)
    * Es ordenada (secuencial)
    * Es indexada

* **tuplas** (tuple): 
    * Son inmutablese (no se pueden cambiar)
    * son ordenadas (secuenciales)
    * son indexadas
    * son mas rapidas que las listas.

* **diccionarios** (dict): 
    * No son ordenados
    * Tiene claves y valores. Estas sirven para re-organizarlo o accesarlo
    
* **sets** (conjuntos set): 
    * no son ordenados
    * son mutables
    * no hay indice

### List comprehension 

La sintaxis de la "comprension" de listas es

`newList = [ *expression*   for    *item*  `in`   *iterable*  `if`  *condition == True*]

* `expression`  es una funcion de uno o varios parametros
* `item` es la variable independiente
* `iterable`  un objeto que sobre el cual se puede iterar, por ejemplo un range, tuple, list, etc.
* `condition`  es un filtro, si el valor es `True` se incluye en la lista, sino se excluye.


```python
myList = [ i**2  for i in range(10) if i%2==1]  # los impares al cuadrado hasta 9
myList
```

```python
myList = [ i**2  for i in range(10) if i%2]
myList
```

## Como saber si un objeto es `container`  o no.

En el ejemplo siguiente verificamos que un arreglo en numpy es container.

```python
import numpy as np
a = np.array([1,1,3])
membersOfa = dir(a)

if membersOfa.count("__contains__"): # count should return 1
    print("a is a container")
```

# Iterators (iteradores)

Un iterador es un objeto que contiene un numero contable de valores. 
El iterador se puede "atravesar" a traves de estos valores. Los iteradores se implementan con el "magic" `__iter_()` y   `__next__()`. 

Comencemos con otros objetos que se llaman `iterables' 

```python
L = ["a","b","c"]
# se puede iterear sobre una lista

for l in L:
    print(l)
```

```python
# una lista es un iterable
# para saberer si un objeto es iterable toca ver que
# dentro del dir(), existe el metodo __iter__()

def isIterable(myObj):
    membersOfObj = dir(myObj)

    if membersOfObj.count("__iter__"):
        print(f"Object  {myObj} is an iterable")
    else:
        print(f"Object {myObj} is not iterable")
    return

isIterable(L)
```

```python
# como saber si un objeto es iterador.
def isIterator(myObj):
    membersOfObj = dir(myObj)

    if membersOfObj.count("__next__"):
        print(f"Object {myObj} is an iterator")
    else:
        print(f"Object {myObj} is not an iterator")
    return

isIterator(L)
```

```python
# vamos a crear nuestra propia clase Range

class Range:

    def __init__(self, start, end, increment):
        self.value = start
        self.end = end
        self.increment = increment
        return

    def __iter__(self):
        return self


    def __next__(self):
        if self.value >= self.end:
            raise StopIteration
        current = self.value
        self.value += self.increment
        return current


# probemoslo
nums = Range(1, 10, 2)


```
