# Multiplicacion de matrices.
Si $A \in \mathbb{C}^{m \times n}$ y $B \in \mathbb{C}^{n \times p}$, entonces $C=AB$ donde $C$ es el producto de $A$ y $B$ si

$$ c_{ij} = \sum_{k=1}^n a_{ik} b_{kj}.$$
$i=1, \cdots, m$, $k=1, \cdots, n$, $j=1, \cdots , p$.

Comentarios:

* El numero de columnas de $A$ debe ser igual al numero de filas de $B$ (en este caso $n$.
* La matriz producto tiene el numero de filas de $A$ y el numero de columnas de $B$.


**Teorema 2.1.1:**: Leyes basicas de las matrices

1. La suma de matrices es **conmutativa**. Es decir
$$ A+B = B+A.$$
Las matrices $A,B$ pueden ser reales co complejas.

**Prueba** (**ojo asi quiero que me hagan la tarea)
Sea $C=A+B$, entonces


$$ c_{ij} = a_{ij} + b_{ij} \quad \text{ por definicion de igualdad de matrices y suma de matrices} $$


$$ c_{ij} = b_{ij} + a_{ij} \quad \text{ por conmutatividad de la suma numeros reales o complejos} $$

$$ C = B+A \quad  \text{por definicion de suma y igualdad de matrides} $$

2. El suma y el producto de matrices es **asociativo**.

$$ A + (B + C) = (A +  B) +  C.$$
$$ A(BC) = (AB) C $$

3. La ley **distributiva**
 $$A(B \pm C) = AB \pm AC  .$$

4. El **producto y transpuesta de transpuestas**:

    * $(AB)^T  = B^T A^T$
    * $(A^T)^T = A$

**Definicion producto por escalar (numero real o complejo)**:
Sea $\alpha$ (recuerden que en este curso, las letras griegas minusculas son escalar). Definimos el producto $B=\alpha A$ como
si $b_{ij} = \alpha a_{ij}$, $\forall i, j$


**Definicion: Producto Hadamard**. Tambien conocido como producto **Schur**. Sean $A, B$ matrices ambas $m \times n$. Definimos
el producto Hadamard como $C= A \circ B$, cuando

$$c_{ij} = a_{ij} b_{ij} .$$
Es un producto componente-por-componente.
Observe que el numero de multiplicaciones en el producto Hadamard es $m n$. Note tambien que las dimensiones de las matrices $A,B$ son las mismas.




**Teorema 2.1.2**: Las siguietes propiedades son del producto Hadamard

* Conmutativa: $$A \circ B = B \circ A$$
* Asociativa: $$A \circ ( B  \circ C) = (A \circ B) \circ C .$$
* Distributiva: $$A \circ ( B  \pm  C) = (A \circ B)  \pm  A \circ C .$$

**Definicion Matriz Adjunta (Adjoint)**: Dada una matriz $A$ la
**conjugada** de $A$ se nota $\overline{A}$ es la matriz
$B = \overline{A}$, tal que $b_{ij} = \overline{a_{ij}}$.
Recuerden que $c=a + \mathrm{i} b$ es un numero complejo con $a,b \in \mathbb{R}$ y $\mathrm{i}= \sqrt{-1}$, $\overline{c}=a - \mathrm{i} b$. Ahora, la matriz **adjunta** es la transpuesta conjugada. Es decir $B$ es la matriz **adjunta** de $A$,
y la notamos $B=A^*$ si

$$A^* = \overline{(A^T)} .$$
Es decir cada elemento de $B$ es

$$b_{ij} = \overline{a_{ji}} .$$

**Definicion producto interno**: Sean $u,v$ vectores en $\mathbb{C}^n$, entonces definimos el producto interno con la ecuacion

$$ \langle u , v \rangle = u^* v.$$

Un ejemplo

$u = \begin{pmatrix}  3 + 2 \mathrm{i} \\  7 +  6 \mathrm{i} \end{pmatrix}$
$v = \begin{pmatrix} 3 -4 \mathrm{i} \\ 5 \mathrm{i} \end{pmatrix}$

El producto interno $\langle u, v \rangle = u^* v $ es

$$ \langle u , v \rangle = \begin{pmatrix} 3 - 2 \mathrm{i} & 7 - 6 \mathrm{i} \end{pmatrix} \begin{pmatrix} 3 -4 \mathrm{i} \\ 5 \mathrm{i} \end{pmatrix} = (3 - 2 \mathrm{i})(3 - 4 \mathrm{i}) + (7 - 6 \mathrm{i})(5 \mathrm{i}) $$

### Propiedades del producto interno.

* Si $u,v$ son reales $\langle u, v \rangle = u \cdot v$
Veamos.  Por definicion de producto de matrices
$u = (u_1, u_2, \cdots, u_n)^T$
$v = (v_1, v_2, \cdots, v_n)^T$

$$ \langle u, v \rangle = \sum_{i=1}^n \overline{u_i} v_i.$$
pero si $u$ es real $u_i = \overline{u_i}$ de forma que

$$ \langle u, v \rangle = \sum_{i=1}^n u_i v_i = u \cdot v$$
Por lo tanto si $u,v$ son reales el producto interno **conmuta**. Si $u,v$ no son reales, les dejo de tarea buscar $u,v$ tales que

$$\langle u, v \rangle \ne \langle v , u \rangle .$$

*  Si $\alpha$ es escalar

\begin{eqnarray}
\langle \alpha u, v \rangle = \overline{\alpha} \langle u , v \rangle  \\
\langle u \pm v, w  \rangle  = \langle u, w \rangle \pm \langle v , w \rangle.
\end{eqnarray}

**Norma (tamanho) de un vector**: La norma de un vector $v$, se nota $\| v \|$ la definimos como


$$\| v \| = \sqrt{ \langle v , v \rangle} .$$
Por ejemplo $v=(a,b)^T$, $a, b \in \mathbb{R}$,  entonces

$$ \| v \| = \sqrt{ \begin{pmatrix} a & b \end{pmatrix} \begin{pmatrix} a \\ b \end{pmatrix}   } =  \sqrt{a^2 + b^2} $$

A esta norma se le conoce como la norma 2 o $\ell_2$.
Hay muchas
[normas de vectores](https://en.wikipedia.org/wiki/Norm_(mathematics)).

Tambien se le conoce a esta norma como la **Euclidea** ( o **Pitagorica**.


Si $\| v \| = 1$ decimos que el vector $v$ es **unitario**.


Lo anterior era para norma de vectores. Tambien hay normas para las matrices. En este curso usamos la norma **Frobenious**.


Esta es: Sea $A$ una matriz $m \times n$, compleja or real.

$$ \| A  \|_F  = \sqrt{ \sum_{i=1}^m \sum_{j=1}^n | a_{ij} |^2}.$$

**Definicion: Matriz identidad**:
La matriz $n \times n$, $I$, se define como la identidad si cumple las propiedades del Kronecer delta


\begin{eqnarray}
\delta_{ij} =
\begin{cases}
1 \quad , \quad \text{si} \quad i = j \\
0 \quad , \quad \text{si} \quad i \ne  j
\end{cases}
\end{eqnarray}

$$
I =
\begin{pmatrix}
1 & 0 & 0 & \cdots  & 0  \\
0 & 1 & 0 & \cdots  &  \vdots   \\
0 & 0 & 1 & \cdots  &  \vdots   \\
\vdots & \vdots & \vdots & \ddots & 0  \\
0 & 0 & 0 & \cdots  &  \vdots  1 \\
\end{pmatrix}
$$

Se propar que si $A$ es $n \times n$ entonces $AI = IA = A$.

Por definicion dijimos que una matriz era **ortogonal** si las
columnas eran ortogonales entre ellas y con norma 1 (ortonormales).
Sera que las filas tambien lo son?


Esto lo vamos a probar. Pero primero probamos que si $A$ es ortogonal entonces
$$ A^* A = I$$
Sea $v_i$ la columna $i$ de $A$

Como se multiplican $C=A^* A$. Primero el elemento $c_{ij}$ que
viene de multiplicar $AB$, es la fila $i$ de $A$ por la columna $j$ de $b$. Ahora bien, la fila $i$ de $A^*$ es la columna $i$ de $A$, estas columas son ortogonales, de forma que

$$ c_{ij} = \langle  v_i, v_j \rangle = \delta_{ij}$$
De forma que $A^* A = I$.
