**Ejemplo 2**. Encuentre los autovalores/vectores de

$$A = \begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix} .$$

Explique el significado de los autovalores/vectores usando un cuadrado de vertices $(1,1), (-1,1), (-1,-1), (1, -1)$ y
un circulo unitario $x^2  + y^2 = 1$.


**Solucion**

$$\det( A - \lambda I) = \det \begin{pmatrix} 1 - \lambda
& 1/2  \\ 1/2 & 1 - \lambda \end{pmatrix} = (1-\lambda)^2 - \frac14 = 0$$

Factorizamos
$$(1 - \lambda - 1/2)(1 - \lambda + 1/2 ) = 0.$$
$$(1/2 - \lambda)(3/2 - \lambda)  $$

Los autovalores son $\lambda_1=3/2$, $\lambda_2=1/2$.


Encontremos los autvectores

* (para $\lambda_1=3/2$)
$$ (-1/2) x + (1/2) y = 0$$
 solucion $x=y$ escogemos

 $$u = \binom{1}{1}$$

* (para $\lambda=1/2$)

$$ \frac12 x + \frac12 y = 0.$$
es decir $x=-y$, escoja el que quiera

$$v = \binom{1}{-1} .$$

Iterpretamos los resultados en un cuadrado.

\begin{eqnarray}
\begin{pmatrix}
1 & 1/2 \\ 1/2 & 1
\end{pmatrix}
\binom{1}{1} = \binom{3/2}{3/2} \\
\begin{pmatrix}
1 & 1/2 \\ 1/2 & 1
\end{pmatrix}
\binom{-1}{1} = \binom{-1/2}{1/2} \\
\begin{pmatrix}
1 & 1/2 \\ 1/2 & 1
\end{pmatrix}
\binom{-1}{-1} = \binom{-3/2}{-3/2} \\
\begin{pmatrix}
1 & 1/2 \\ 1/2 & 1
\end{pmatrix}
\binom{1}{-1} = \binom{1/2}{-1/2} \\
\end{eqnarray}

En palabras: El vector $u$ se **estira** una cantidad de $3/2$,
el vector $v$ se **encoje** $1/2$.
Leer en las notas de clase, la parte cuantitativa de esta afirmacion. Pagina 33.

Veamos el efecto de la matriz sobre los autvectores/valores.

$$ \| u  \| = \| v \| = \sqrt{2}.  $$
Ahora $u'=Au$, $v'=Av$,

$$ \| u' \|  = \sqrt{ (3/2)^2 + (3/2)^2 } = \sqrt{9/2}=3/\sqrt{2} $$

$$ \| v' \|  = \sqrt{ (1/2)^2 + (1/2)^2 } = 1/\sqrt{2}.$$

Veamos los cocientes de las normas de salida sobre las normas de entrada.

$$ \frac{\| u' \|}{\| u \|} = (3/\sqrt{2})/\sqrt{2} = 3/2 $$
$$ \frac{\| v' \|}{\| v \|} = (1/\sqrt{2})/\sqrt{2} = 1/2 $$

Pero $\lambda_1=3/2$ y $\lambda_2=1/2$.

Esto se puede explicar de la siguiente forma

$$\frac{ \| u' \|}{\| u \|} = \frac{ \| A u \| }{\| u \|} =\frac{ \| \lambda_1 u \|}{\| u \|} = (3/2) \frac{\| u \|}{ \| u \|}=3/2  .$$

$$\frac{ \| v' \|}{\| v \|} = \frac{ \| A v \| }{\| v \|} =\frac{ \| \lambda_2 v \|}{\| u \|} = (1/2) \frac{\| v \|}{ \| v \|}=1/2  .$$

**Ejemplo 3**: Encuentre los autovalores/vectores de la matriz

$$A = \begin{pmatrix}  1 & 0 \\ 0 & 0 \end{pmatrix} .$$
Explique el resultado de $A$ aplicado al cuadrado de vertices
$(1,1), (-1,1), (-1, -1), (1, -1)$.

**Solucion**: El polinomio caracteristico.

$$ p(\lambda) = \det \begin{pmatrix} 1 - \lambda & 0 \\ 0 & -\lambda \end{pmatrix} = -\lambda(1 - \lambda) = 0.$$


Los autovalores son $\lambda_1 = 1$, $\lambda_2 = 0$.


Tarea: (voluntaria)
Pruebe que el autovector correespondiente a 1 es $u=(1,0)^T$
y el autovector correspondiente a $0$ es $v=(0,1)^T$
Apliquemos la matriz al cuadrado.

\begin{eqnarray}
\begin{pmatrix}
1 & 0 \\ 0 & 0
\end{pmatrix} \binom{1}{1} &=& \binom{1}{0} \\
\begin{pmatrix}
1 & 0 \\ 0 & 0
\end{pmatrix} \binom{-1}{1} &=& \binom{-1}{0} \\
\begin{pmatrix}
1 & 0 \\ 0 & 0
\end{pmatrix} \binom{-1}{-1} &=& \binom{-1}{0} \\
\begin{pmatrix}
1 & 0 \\ 0 & 0
\end{pmatrix} \binom{1}{-1} &=& \binom{1}{0}.
\end{eqnarray}

# Teorema espectral.

**Teorema 2.2.2**: Si $A$ es auto-adjunta, ($A^* = A$)entonces sus autovectores, correspondientes a autovalores distintos, son ortogonales y sus autovalres son reales.

**Prueba**: Probemos primero que los autovalores son reales.

Asuma $\lambda_j$ es un autovalor de $A$ con autovector $u_j$.
Sin perdida de generalidad podemos asumir que $u_j$ es unitario,
estudiamos la forma **quadratica** $\langle A u_j, u_j \rangle$

$$\langle A u_j , u_j \rangle = \langle \lambda_j u_j , u_j \rangle =  \overline{\lambda_j} \langle u_j, u_j \rangle = \overline{\lambda_j}   .$$

Ahora veamos que $\langle A x, y \rangle = \langle x, A^* y \rangle $

Prueba:

$$\langle A x, y \rangle = (A x)^* y = x^* A^* y = \langle x, A^* b \rangle
=\langle u_j, A u_j \rangle = \langle u_j, \lambda_j u_j \rangle = \lambda_j \langle u_j, u_j \rangle = \lambda_j  .$$

Si $\overline{\lambda_j} = \lambda_j $
es por que $\lambda_j$ es real .

($a+ \mathrm{i}b = a - \mathrm{i}b$, entonces $b=0$)



De forma que
$$ \overline{\lambda_j} = \langle A u_j, u_j \rangle =\langle u_j, A^* u_j \rangle = \langle u_j .$$

Ahora probamos que si $u,v$ son dos autovectores correspondientes a $\lambda_1 \ne \lambda_2$ entonces


$$\langle A u , v \rangle = \langle \lambda_1 u, v \rangle = \lambda_1 \langle u, v \rangle  \tag{1} .$$

Ahora, usando $A = A^*$,

$$\langle Au, v \rangle = \langle u, A v \rangle = \langle u, \lambda_2 v \rangle = \lambda_2 \langle u, v \rangle . \tag{2}$$

Restando la (1) de la (2)

$$(\lambda_1 - \lambda_2) \langle u, v \rangle = 0 .$$
y como $\lambda_1 - \lambda_2 \ne 0$, entonces

$$ \langle u, v \rangle = 0 $$
y $u$ es ortogonal a $v$.

Si todos los autovalores son distintos, entonces todos los autovectores son ortogonales y descomponemos el espacio $\mathbb{C}^n$ en $n$ vectores mutuamente ortogonales.
Sean $\{ x_1, x_2, \cdots, x_n \}$ todos los autovectores de $A$.
Es decir $A x_i = \lambda_i x_i$, $i=1,2, \cdots, n$
Ahora como los $x_i$ son ortogonales entre si

$$x_i^* x_j = \delta_{ij} .$$

Entonces

$$ x_i^* A  x_j = x_i^* (\lambda_j x_j) = \lambda_j x_i^* x_j = \lambda_j \delta_{ij} .$$

En forma matricial

$$Q^* A Q = \Lambda . \tag{3} $$
Donde $Q$ es una matriz cuyas columnas son los $x_i$ y
$\Lambda$ es una matriz diagonal con todos los autovalores.








Esto pasa si todos los autovalores son distintos, pero **tambien** pasa si hay repetidos. La prueba de esto para autovalores repeditos la voy a omitir pero esta en las notas.

En la ecuacion (3), multipliquemos a izquierda por $Q$ y a derecha por $Q^*$, entonces

$$ Q Q^* A Q Q^* =  Q \Lambda Q^* .$$
$$ I A I  =  Q \Lambda Q^* .$$
$$  A   =  Q \Lambda Q^* . \tag{4} $$

El teorema espectral para matrices autoadjuntas dice que si $A$ es autoadjunta exte una matriz $Q$ (cuyas columnas son los autovectores de $A$), y una matriz diagonal $\Lambda$ que son los autovalres de $A$ tal que se cumple le ecuacion (4)


Veamos la importancia de este teorema:

La forma cuadratica. Sea $x \in \mathbb{R}^n$, $A \in \mathbb{R}^{n \times n}$, entonces, $A$ es simetrica.


$$ \langle x, Ax \rangle =  x^T A x = \sum_{i=1}^n \sum_{j=1}^n a_{ij} x_i x_j  .$$
Cuantos terminos tiene esa suma? Tiene $n^2$ y cada termino
tiene 2 multiplicaciones, o sea que hay $2 n^2$ multiplicaciones. Sumas $n^2$.

Veamos como se puede reducir este calculo con la diagonalizacion.
Si $A = Q \Lambda Q^T$, entonces

$$x^T A x = x^T Q \Lambda Q^T x . \tag{5}$$
sea $y=Q^T x$, entonces (5) se puede escribir como

$$ x^T A x = y^T \Lambda y = \sum_{i=1}^n \lambda_i y_i^2.$$
Reducimos el problema (luego de representar $x$ en la nueva base como $y$) en $n$ sumas y $n$ cuadrados, $n$ multiplicaciones.


Otras aplicaciones de esto es representar llevar las conicas a las formas canonicas.

* Ellipse ($ax^2 + b xy + cy^2$) $\implies$  $\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$
* Hiperobola ($ax^2 + b xy - cy^2$) $\implies$  $\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1$
* Parabola ....


Exponenciacion de una matriz.

$A = Q \Lambda Q^T$, $\Lambda$ es diagonal.

\begin{eqnarray}
A^2 &=& (Q \Lambda Q^T) (Q \Lambda Q^T)  \\
&=& (Q \Lambda ( Q^T Q) \Lambda Q^T)  \\
&=& (Q \Lambda  I  \Lambda Q^T)  \\
&=& (Q \Lambda    \Lambda Q^T)  \\
&=& (Q \Lambda^2  Q^T)  \\
\end{eqnarray}
Y asi sucesivamente

$$A^n = Q \Lambda^n Q^T .$$

Donde

$$\Lambda^n = \begin{pmatrix}
\lambda_1^n & 0 & \cdots   &  0   \\
0 & \lambda_2^n  & \ddots  &  \vdots   \\
\vdots  & \vdots  & \ddots   &  0  \\
0 & 0 & \cdots  &  \lambda^n   \\
\end{pmatrix}
.$$
