# La Pseudoinversa.
## introduccion.
Sabemos que si $A$ es $n \times  n$ posiblemente tenga inversa, pero si $A$ es $m \times n$, donde $m \ne n$, no tiene inversa en ningun caso. No existe la inversa. Queremos generalizar el concepto de inversa para matrices no cuadradas.

Como vamos a usar los conceptos de espacio nulo y rango los repasamos.

Sea $A$ una matriz. El espacio nulo de $A$ que se noto $\mathcal{N}(A)$ es el espacio de todos los $x$ tales que $Ax = 0$.
Tiene una propidad muy importante. Si $x_p$ es una solucion
de $Ax=b$, entonces $x_p + x_N$, con $x_N$ en el espacio tambien es solucion. Por que ?

$$A (x_p + x_N)= A x_p + A x_N  .$$
pero $A x_N=0$, por $x_N \in \mathcal{N}(A)$, etonces

$$A (x_p + x_N)= A x_p + A x_N  = A x_p = b.$$
Cuando el espacio nulo es el trival $\{ 0 \}$ la solucion
es unica.

El otro espacio importante es el rango. Para $A$ una matriz,
el rango de $A$, $\mathcal{R}(A)$ es el espacio columna, es decir, todas las combinaciones lineales de las columnas.

Decir que

$$ Ax = b.$$
es lo mismo que decir que $b \in \mathcal{R}(A)$.


Recuerden que el problema principal que , por el momento, queremos resolver es $Ax = b$.


Listamos el siguiente arbol de decision.

* Si $b \in \mathbb{R}(A)$ entonces la solucion existe. Luego de aca existen dos posibilidades.
    1. Si $\mathcal{N}=\{ 0 \}$, entoncens  la solucion es unica y existe $A^{-1}$, de forma que
    $$ x = A^{-1} b$$.
    En este caso la **pseudoniversa** es $A^+ = A^{-1}$.
    2. Si
 $\mathcal{N} \ne \{ 0 \}$ entonces la solucion no es unica. Entonces podemos encontrar la mas pequenha de todas, $\min \| x \|$ tales que $A x = b$. En este caso
$x= A^+ b$ donde $A^+$ es la pseudoinversa.

    Veamos un ejemplo. Resuelva
    $$ 2 x + 3 y = 5.$$
    Escribamosla en forma $Ax = b$

 $$ \begin{pmatrix} 2 & 3 \end{pmatrix} \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} =  3.$$
Aca la matriz $A=(2  \quad 3)$.
Quien es el espacio nulo de $A$? Para esto resolvemos el sistema $Ax=0$,

$$ 2 x_1 + 3 x_2 = 0 .$$
Existen infinitas soluciones, por ejemplo $x=-3, y=2$, y todos
los multiplos de esta son soluciones

$$\mathcal{N}(A) = \alpha \binom{-3}{2}  .$$

En el arbol que venimos trabajando

* $b \in \mathcal{R}(A)$.
Esto quiere decir que $b \ne Ax$ para ningun $x$. Es decir
$$ Ax = b$$ **no tiene solucion**.

Toca **proyectar** $x$ al rango de $A$ y esto se hace con las ecuaciones normales.

$$ Ax  = b.$$
no tiene solucion, pero multiplicamos a ambos lados por $A^T$

$$ A^T A x= A^T b$$
estas son las ecuaciones normales y tiene solucion.

De nuevo de aca salen dos ramas.

    1. Primero que
$\mathcal{N} = \{ 0 \}$
Esto implica que la solucion es **unica** y $A^T A$ tiene inversa
Es decir

$$ x= (A^T A)^{-1} A^T b  $$
y esta pseudoslucion y la matriz $(A^T A)^{-1} A^T es la
pseudoinversa. Es decir

$$A^+ = (A^T A)^{-1} A^T .$$

2. Ultimo caso
$\mathcal{N} \ne \{ 0 \}$. No existe solucion unica. Se pueden
hacer dos cosas. Regularizacion y limite, o lo que dice Keener
es hallar el minimo $\|x \|$ de todas las infinitas soluciones.
Esto se halla garantizando que $\langle x, y \rangle = 0$,
$\forall y \in \mathcal{N} $.

En este caso la pseudoinversa es

$$A^+ = \lim_{\alpha \to 0} (A^T A + \lambda I)^{-1} A^T .$$

Que relacion tiene esto con la pseudo-inversa de Moore-Penrose?

Si, la pseudoinversa que estamos mostrando aca es la misma de Moore-Penrose. La definicion que yo use , para mi, es mas intuitiva.

La defincion de pseudoinversa de Keener es la siguiente:

Sea $A$ una matrix. Si existe una matriz $A^+$ (la pseudoinversa) tal que se cumplan estas dos condiciones

1. $$A^* A x = A^* b  \quad , \quad \text{ecuaciones normales} $$
2. $$\langle x, w \rangle  =0 \quad \text{       para todo $w$, que esta en el espacio nulo de $A$} $$


Keener (referencia en las notas de clase) presenta como 4 algoritmos para hallar la pseudoinversa (descomposicion $LU$, eliminacion Gaussiana, SVD.

Veamos como hallar la pseudoinversa usando SVD.

Reuerde que cualquier matriz $A$,  se puede escribir como

$$A = U \Sigma V^* .$$
donde $U,V$ tienen los autvectores de $A A^*$ y $A^* A$ respectivamente y $\Sigma$ tiene los valores singulares, las
raices cuadradas de los autovectores de $A^* A$.


$\Sigma$ tiene la siguiente estructura


$$\Sigma = \begin{pmatrix}  \Sigma_r & | & 0 \\ ---&---&--- \\ 0 & | & 0 \end{pmatrix}  .$$
Si $A$ es $m \times n$ entonces $\Sigma$ tambien es $m \times n$.

Recuerden que como las matrices $U,V$ son otrogonales
$$U^{-1} = U^* \quad , \quad V^{-1}=V^* $$
Ahora bien la pseudoinversa de $\Sigma$ es una matriz de orden
$n \times m$.  La matriz $\Sigma^+$ tiene los inversos de $\sigma_i$, es decir $\sigma_i^{-1}$ en la diagonal y ceros en cualquier otro lado.

\begin{eqnarray*}
  \Sigma^+ = \begin{pmatrix}
    \sigma_1^{-1} & 0 & \cdots & \cdots & \cdots & \cdots & 0 \\
    0 &  \sigma_2^{-1} & \ddots &  &  &  & \vdots \\
    \vdots & \ddots &  \ddots & \ddots &  &  & \vdots \\
    \vdots &  & \ddots & \sigma_r^{-1} & \ddots &  & \vdots  \\
    \vdots &  &  & \ddots &  0 & \ddots  & \vdots  \\
    \vdots &  &  &  &  \ddots & \ddots   & \vdots \\
    \vdots &  &  &  &  & \ddots  & \vdots \\
    \vdots &  &  &   &  &  & \vdots \\
    0 & \cdots & \cdots & \cdots & \cdots & \cdots & 0 \\
  \end{pmatrix} \,.
\end{eqnarray*}





En notacion simplificada, la matriz $\Sigma^+$
Para nosotros $r$ es el numero de valores singulares que no son 0.

\begin{eqnarray}
\sigma_{ij}^+ =
\begin{cases}
\delta_{ij} \sigma_i^{-1} \quad , \quad i \le r \\
0 \quad , \quad \text{de otro modo}
\end{cases}
\end{eqnarray}
La pseudoinversa de $A$ es entonces

$$A^+ = V \Sigma^+ U^* .$$

La pseudosolucion es

$$x^+ = A^+ b = V \Sigma^+ U^* b .$$
