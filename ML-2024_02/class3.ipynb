**Definicion: Inversa de una matriz**
Sea $A$ un matriz cuadrada $n \times n$. La inversa de $A$ existe, y la llamamos $B$ si $AB = BA = I$. Notamos
$B=A^{-1}$.  Decimos que $A$ tiene inversa, es invertible, o es **no-singular**. Si $A$ no tiene inversa se llama **singular**.

En la clase anterior probamos que $A^* A = I$,
ahora vamos a probar que $A A^* = I$ y que $A^* = A^{-1}$ cuando
$A$ es ortogonal.

Antes de probar esto probemos un Lema

**Lema**:
1. Si $A$ y $B$ son matrices $n \times n$, entonces
$\det(AB) = \det(A) \det(B)$.
2. Asumamos que $A$ es $n \times n$ y $\det(A) \ne 0$, entonces $A$ tiene inversa.

**Teorema**:
Si $AB=I$, $A,B$ son cuadradas, entonces $BA=I$. En palabras, si es inversa a derecha es inversa a izquierda, pero la matriz $A$ debe ser cuadrada.

**Prueba**:
Si $AB=I$, entonces

$$ B = B(AB) = (BA)B. \tag{1} $$
como $\det(I)=1$, entonces $\det(AB) = \det(A) \det(B)= 1$.
De forma que $\det(B) \ne 0$. O sea que $B$ tiene inversa (punto 2 del lema). Entonces multipliquemos por $B^{-1}$ a derecha en la ecuacion  (1)


$$  B B^{-1}  = (BA)B B^{-1}. $$

$$  I  = (BA). $$
entonces $I=BA$.

Ahora si, del teorema de la ultima clase $A A^*=I$, entonces
$A^* A =I$. En palabras: Si es ortogonal por filas, tambien lo es por columnas.  

**Teorema**: Si $A,B$ son no-singulares, entonces $AB$ es no-singular y $(AB)^{-1}=B^{-1} A^{-1}$.

**Prueba**:
Como $A,B$ son invertibles, existen $A^{-1}$, $B^{-1}$, y

$$(B^{-1} A^{-1}) (AB)= (B^{-1})(A^{-1} A) B = B^{-1} I B =
B^{-1} B = I.$$
Entonces, por definicion de inversa $B^{-1} A^{-1}$ es la inversa
de $AB$.


**Teorema**: Si $A$ es **ortogonal** $A \in \mathbb{C}^{n \times n}$, entonces $\| A x \| = \| x \| $

En palabras esto dice que la matriz $A$ es **isometrica**.

**Prueba**:

$$ \| A x \|^2 = (A x)^* Ax = x^* A^* A x = x^* I x = x^* x = \| x \|^2 .$$

Si $A$ es real, cambie $A^*$ por $A^T$.

**Definicion** (espacios de rango y nulo). Dada una matriz $m \times n$, definimos el espacio **rango** o espacio **columna** como el espacio generado por todas las columnas de $A$.
La notacion del espacio rango es $\mathcal{R}(A)$. Es decir,
si las columnas de $A$ se notan como $A_i$, $i=1,2, n$, entonces
el rango son todos los vectores de la forma

$$ \sum_{i=1}^n \alpha_i A_i .$$
donde $\alpha$ es cualquier escalar (real o complejo).

El espacio **nulo** es el espacio de $A$ de todos los $x \in \mathbb{R}^n$, tales que $A x = 0$.


##Autovalores/ Autovectores (Eigenvalues/Eigenvectors)
Si para una matriz cuadrada $A \in \mathbb{C}^{n \times n}$
existe un vector $u \ne 0  \in \mathbb{C}^n$ tal que

$$ A u = \lambda u \tag{2} $$
entonces decimos que $\lambda$ es un **autovalor** de la matriz $A$ y que $u$ es un **autovector** correspondiente a $\lambda$ de la matriz $A$.

Como se hallan los autovalores/vectores?
Veamos
De la ecuacion (2)



$$ A u - \lambda I u = (A - \lambda I)u = 0$$

Del algebra lineal $C u=0$ y $ u \ne 0$, entonces el
determinate de $C$ es 0, $C$ es cuadrada.

Entonces tenemos que

$$p(\lambda) = \det ( A - \lambda I) = 0 .$$

Antes de ver ejemplos, un teorema importante

**Teorema** (existencia de autovalores). Caulquier matriz $\mathbb{C}^{n \times n}$ para $n \ge 1$ **tiene que tener**
por lo menos un autovalor.

**Ejemplo 2**. Encuentre los autovalores/vectores de

$$A = \begin{pmatrix} 1 & 1/2 \\ 1/2 & 1 \end{pmatrix} .$$

Explique el significado de los autovalores/vectores usando un cuadrado de vertices $(1,1), (-1,1), (-1,-1), (1, -1)$ y
un circulo unitario $x^2  + y^2 = 1$.


**Solucion**

$$\det( A - \lambda I) = \det \begin{pmatrix} 1 - \lambda
& 1/2  \\ 1/2 & 1 - \lambda \end{pmatrix} = (1-\lambda)^2 - \frac14 = 0$$

Factorizamos
$$(1 - \lambda - 1/2)(1 - \lambda + 1/2 ) = 0.$$
$$(1/2 - \lambda)(3/2 - \lambda)  $$

Los autovalores son $\lambda_1=3/2$, $\lambda_2=1/2$.


Encontremos los autvectores

* (para $\lambda_1=3/2$)
$$ (-1/2) x + (1/2) y = 0$$
    solucion $x=y$ escogemos

    $$u = \binom{1}{1}$$

* (para $\lambda=1/2$)

$$ \frac12 x + \frac12 y = 0.$$
es decir $x=-y$, escoja el que quiera

$$v = \binom{1}{-1} .$$

Iterpretamos los resultados en un cuadrado.
