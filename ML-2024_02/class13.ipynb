## Regresion polinomial en polinomios multivariados
Por ejemplo piense que quiere vender la casa y tiene dos parametros (el tamanho y el numero de habitaciones). Piense que la funcion de regresion es polinomial (en dos variables) y cuadratica.

La hipotesis es:

$$h(\theta_0, \theta_1, \theta_2, \theta_3, \theta_4, \theta_5, x_1, x_2 )=
\theta_0 + \theta_1 x_1 + \theta_2 x_2 + \theta_3 x_1 x_2 + \theta_4 x_1^2 + \theta_5  x_2^2 $$

Vamos a construir la matriz.

$$ \theta_0 + \theta_1 x_{11} + \theta_2 x_{12} + \theta_3 x_{11} x_{12} + \theta_4 x_{11}^2 + \theta_5  x_{12}^2 $$

$$ \theta_0 + \theta_1 x_{21} + \theta_2 x_{22} + \theta_3 x_{21} x_{22} + \theta_4 x_{21}^2 + \theta_5  x_{22}^2 $$

$$ \vdots $$

$$ \theta_0 + \theta_1 x_{m1} + \theta_2 x_{m2} + \theta_3 x_{m1} x_{m2} + \theta_4 x_{m1}^2 + \theta_5  x_{m2}^2 $$

La matriz queda


$$ X = \begin{pmatrix}
1 & x_{11} & x_{12} & x_{11} x_{12} & x_{11}^2 & x_{12}^2 \\
1 & x_{21} & x_{22} & x_{11} x_{22} & x_{21}^2 & x_{22}^2 \\
\vdots & \vdots & \vdots & \vdots & \vdots & \vdots   \\
\vdots & \vdots & \vdots & \vdots & \vdots & \vdots   \\
\vdots & \vdots & \vdots & \vdots & \vdots & \vdots   \\
1 & x_{m1} & x_{m2} & x_{11} x_{m2} & x_{m1}^2 & x_{m2}^2 \\
\end{pmatrix} $$
Debemos resolver $X \theta = y$

Donde

$$\theta = \begin{pmatrix}  \theta_0 \\ \theta_1 \\ \theta_2 \\
\theta_3 \\ \theta_4 \\ \theta_5 \end{pmatrix} $$

$$y = \begin{pmatrix}  y_1   \\ \vdots  \\ \vdots  \\ y_m \end{pmatrix}  $$

Sigue siendo el mismo problema que hemos venido solucionando:

$$ X^T X \theta = X^T y  $$
o si se necesita regulariacion

$$ (X^T X  + \lambda X) \theta = X^T y  $$

## Overfitting/Underfitting

Vamos a pasar a una regresion lineal mas general que todas estas.
Pero antes voy a discutir dos conceptos muy importantes

* Underfitting (sub-ajuste)
    * Es quedarse corto
    * Es simplificar demasiado el problema. Por ejemplo el problema es intrinsicamente cuadratico y se resuelve como lineal.
    * El error es grande. No se aprendio lo suficiente (perdio el examen).



* Overfitting (sobre-ajuste)
    * La solucion es **demasiado buena** para ser verdad.
    * En el entrenamiento los resultados son perfectos.
    * En la prueba "se raja" .

A mi me gusta construir el siguiente ejemplo. Usted a un ninho le ensenha los animales con juego de loteria (dibujos en tarjetas). Despues de varias horas el ninho memorizo los nombres de los animales. Luego se consigue otra loteria donde los animales estan en posiciones distintas (acostodos, sentados, en el agua) y de pronto con colores ligeramente distintos. Hay dos situaciones

*  Usted evalua al ninho con las cartas originales. El ninho sasa 5.0
*  Usted evalua al ninho con las cartas nuevaes. El ninho sasa 2.4
Esto es **overfitting**.

Usted le ensenha al ninho que 2 naranjas + 3 naranjas son 5 naranjas.
Despues le pregunta cuando es 2 mangos + 3 mangos. El ninho dice que solo se sabe el problema con "naranjas"

**overfitting**: Es la incapacidad del sistema (sea ninho o computador)
de generalizar. Es decir extender la solucion cuando son datos **nuevos**.

Cuando ocurre overfitting:
* Muchos parametros (features). Por ejemplo un polinomio de grado 15 no es acertado.

* No hay regularizacion

Cuando ocurre underfitting
* Muy pocos parametros. Un polinomio lineal no es acertado
* Mucha regularizacion $\lambda \gg 1$

Si hay overfitting.
* Reducir el numero de "features"
* aplicar regularizacion
* jugar con distintas combinaciones de datos

## k-fold crossvalidation.
Se parten los datos en $k$ grupos.

* Se entrena con los grupos $2, 3, \cdots, k$. Se prueba con el grupo 1.
* Se entrena con los grupos $1, 3, \cdots, k$. Se prueba con el grupo 2.
* ... $\vdots $
* Se entrena con $1, 2, \cdots, j-1, j+1, \cdots k$ y se entrena con el grupo $j$
* Se entrena con $1,2, \cdots, k-1$ y se prueba con $k$.
Por que es bueno?
Se usan **todos** los datos para entrenar y para validar.

En esencia $k-$fold cross-validation didive los datos en $k$ grupos y

1. Entrena el modelo con $k-1$ y prueba con el que se quedo por fuera.
2. Rota el experimento.
3. Calcula el error (RMSE:Root Mean Square Error)
4. Luego los $k$ experimentos reporta el promedio de los errores.

### Introduccion
Para poder comparar, recordamos los **metodos tradicionales**.

Vamos a usar letras maysculas $\Theta$, $Y$. Por que?


Tenemos variables cateogoricas. El problem

$$h(\Theta) = X \Theta = Y $$

$Y$ es categoria, puede ser

$$ \text{verdadero} = \binom{1}{0}  $$
$$ \text{falso} = \binom{0}{1}  $$

El problema que queremos resolver es

$$\min_{sobre \; \Theta} J(\Theta) = \min_{\Theta} \frac12 \| X \Theta - Y \|^2$$
