* Datos de entrada $x_i$, $i=1,2, \cdots, m$
* La parte lineal ($z$)
Para cada $i=1,2, \cdots, p$ ($p$ el numero de nodos de la segunda capa.

$$z_i^{(2)} = x^T \Theta_i^{(T)} \quad, i=1,2 \cdots, p $$

De forma indical (granulada)
Digamos que $\Theta$ $n$ columnas

Ciclo 1
$$z_i^{(2)} = \sum_{k=0}^n x_k \theta_{ki}^{(1)}$$

* Activacion para la segunda capa

$$a_{i}^{(2)} = \frac{1}{1 + \mathrm{e}^{-z_{i}^{(2)}}} , i=1,2, \cdots, p $$

Ciclo 2
$$z_i^{(3)} = \sum_{k=0}^n a_k^{(2)} \theta_{ki}^{(2)}, i=1,2,3, ..$$

Esto se puede escribir en forma mas condensada

$$z_i^{(3)} = (a^{(2)})^T \Theta_i^{(2)} $$
* Activacion para la segunda capa

Ignorando el numero de capas de la figura, pensemos que son 3 capas. Ya ultima es esta, la 3
$$h_i(\Theta, x) = a_{i}^{(3)} = \frac{1}{1 + \mathrm{e}^{-z_{i}^{(3)}}} , i=1,2 $$


**Nota**: Un problema lineal es $Ax =b$ (regresion lineal)

Que pasa si todas las activaciones son la identidad.
Entonces una red neuronal de tres capas se modelo como siguie
* datos $x$
* Primera capa $\Theta^{(1)} x$
* Segunda capa $ \Theta^{(1)} I \Theta^{(2)} x$
* Tercerca capa $\Theta^{(1)} I  \Theta^{(2)} I \Theta^{(3)} x$

Llamemos
$A =\Theta^{(1)} \Theta^{(2)} \Theta^{(3)}$
la salida la llamos $b$

Entonces tenemos el problema
$A x= b$

$x^T A^T = b^T$


Para multiclase ($K \ge 3$) se tiene una activacion llamada
la

* Para $(x_1, x_2) = (1,1)$:

$$z_{11}^{(2)} = -10 + 20 x_1 + 20 x_2 = 30 $$
$$a_{11}^{(2)} = S(z_{11}^{(2)}) =\frac{1}{1 + \mathrm{e}^{-30}} \approx 1 $$

$$z_{12}^{(2)} = 30 - 20 x_1 - 20 x_2 = -10$$

$$a_{12}^{(2)} = S(z_{12}^{(2)}) =\frac{1}{1 + \mathrm{e}^{10}} \approx 0. $$

En la ultima capa

$$z_{11}^{(3)} = -30+20 a_{11}^{(2)} + 20 a_{12}^{(2)} \approx -10 $$

La activacion es

$$h(\Theta, x^{(1)}) = S(z_{11}^{(3)}) = \frac{1}{1 + \mathrm{e}^{10}} = 0  $$



* Para $(x_1, x_2) = (1,0)$:

$$z_{21}^{(2)} = -10 + 20 (1) + 20 (0) = 10 $$
$$a_{21}^{(2)} = S(z_{11}^{(2)}) =\frac{1}{1 + \mathrm{e}^{-10}} \approx 1 $$

$$z_{22}^{(2)} = 30 - 20( 1) - 20 0 = 10$$

$$a_{22}^{(2)} = S(z_{12}^{(2)}) =\frac{1}{1 + \mathrm{e}^{-10}} \approx 1. $$

En la ultima capa

$$z_{21}^{(3)} = -30 + 20 a_{21}^{(2)}  + 20 a_{22}^{(2)}= -30 + 20 (1) + 20 (1) = 10  $$

La activacion es

$$h(\Theta, x^{(1)}) = S(z_{11}^{(3)}) = \frac{1}{1 + \mathrm{e}^{-10}} = 1  $$



* Para $(x_1, x_2) = (0,1)$:

$$z_{31}^{(2)} = -10 + 20 (0) + 20 (1) = 10 $$
$$a_{31}^{(2)} = S(z_{11}^{(2)}) =\frac{1}{1 + \mathrm{e}^{-10}} \approx 1 $$

$$z_{32}^{(2)} = 30 - 20( 0) - 20( 1) = 10$$

$$a_{22}^{(2)} = S(z_{12}^{(2)}) =\frac{1}{1 + \mathrm{e}^{-10}} \approx 1. $$

En la ultima capa

$$z_{31}^{(3)} = -30 + 20 a_{31}^{(2)}  + 20 a_{32}^{(2)}= -30 + 20 (1) + 20 (1) = 10  $$

La activacion es

$$h(\Theta, x^{(1)}) = S(z_{11}^{(3)}) = \frac{1}{1 + \mathrm{e}^{-10}} = 1  $$





* Para $(x_1, x_2) = (0,0)$:

$$z_{41}^{(2)} = -10 + 20 (0) + 20 (0) = -10 $$
$$a_{41}^{(2)} = S(z_{11}^{(2)}) =\frac{1}{1 + \mathrm{e}^{10}} \approx 0 $$

$$z_{42}^{(2)} = 30 - 20( 0) - 20 0 = 30$$

$$a_{42}^{(2)} = S(z_{12}^{(2)}) =\frac{1}{1 + \mathrm{e}^{-30}} \approx 1. $$

En la ultima capa

$$z_{41}^{(3)} = -30 + 20 a_{41}^{(2)}  + 20 a_{42}^{(2)}= -30 + 20 (0) + 20 (1) = -10  $$

La activacion es

$$h(\Theta, x^{(1)}) = S(z_{11}^{(3)}) = \frac{1}{1 + \mathrm{e}^{10}} = 0  $$
