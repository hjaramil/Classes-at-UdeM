Se tiene que.

1.
$$ A B = \left [ A B_1 | AB_2 | \cdots A B_p  \right ] \tag{1} $$
donde $B_i$ es la columna $i$ de $B$. En palabras "multiplicar una matriz por otra es multiplicar la matriz por cada una de sus columnas". Esto importante por que si solo queremos la columna $i$ del producto, es suficiente con calcular $A B_i$.

2.
$$ Ax = \sum_{j=1}^n x_j A_j  \tag{2} .$$
donde $A_j$ es la columna $j$ de $A$, $j=1, \cdots, n$.
Esto le que dice es que $Ax$ es una **combinacion lineal** de las columnas de $A$. Es decir que $Ax \in \mathcal{R}(A)$.

3. Si $p \ge n$,

$$A \Lambda = \left [ \lambda_1 A_1 | \lambda_2 A_2 | \cdots | \lambda_n A_n | 0_{n+1} | \cdots 0_p   \right ] \tag{3} .$$
donde agregamos columnas de ceros para rellenar las faltantes desde $n$ hasta $p$. Estas son $p-n$ columnas de ceros. Si $n>p$.

$$A \Lambda = \left [ \lambda_1 A_1 | \lambda_2 A_2 | \cdots | \lambda_p A_p  \right ] \tag{4} .$$

**Teorema de diagonalizacion**: Sea $A$ una matriz diagonalizable (que los autovectores son linealmente independienes), entonces existe una matriz $\Lambda$ con autovalores de $A$ y una matriz $P$ cuyas columnas son los autovectores de $A$, tal que

Probamos que

$$A^k = P \Lambda^k P^{-1} .$$

Probemos esto. Prueba por induccion. En que consiste?
Piense en el efecto domino.  Quiero tumbar todas las fichas.

1. Tumbo la primera (si no se cae la primera no pasa nada. No cae ninguna).
2. Dos fichas deben estar mas cerca que la altura de las mismas. Es decir, si cae la ficha $k$ cae la ficha $k+1$.

Matematicamente el principio de induccion dice que se prueba para $k=1$,
luego que si es valida para $k$ implica que es valida para $k+1$, es valida para todo $k=1,2, \cdots \infty$.

Pricipio de induccion matematica:

1. Se cumple para $k=1$,
Si $k=1$ se tiene
$$A^1 = P \Lambda^1 P^{-1} .$$
esto es  la (9) que es verdad.

2. Se supone que es valida para $k$. Es decir, asumimos

$$A^k = P \Lambda^k P^{-1}  \tag{9b} .$$

Debemos probar que es valido para $k+1$,  es decir, debemos probar

$$A^{k+1} = P \Lambda^{k+1} P^{-1} \tag{10}.$$
Probemos (10)

\begin{eqnarray}
A^{k+1} &=& A^k A \\
&=& (P \Lambda^k P^{-1}) (P \Lambda P^{-1}) \quad \text{por 9b y 9} \\
&=& (P \Lambda^k) (P^{-1} P) ( \Lambda P^{-1}) \quad \text{ley asociativa}  \\
&=& P \Lambda^k I \Lambda P^{-1} \quad \text{pues  $P P^{-1} = I$} \\
&=& P \Lambda^k \Lambda P^{-1}  \\
&=& P \Lambda^{k+1} P^{-1} .
\end{eqnarray}
Esto es (10) que era lo que tocaba probar.

# Singular Value Decomposition (Descomposicion en valores singulares).
Tal vez el teorema mas importante del algebra lineal.
Es muy general, es valido para **todas**.

**Teorema:** (SVD). Dada una matriz $A \in \mathbb{C}^{m \times n}$, existen matrices ortogonales $U \in \mathbb{C}^{m \times m}$, $V \in \mathbb{C}^{n \times n}$, y una matriz diagonal $\Sigma$ de orden $m \times n$ de valores reales no negativos, tales que

$$ A = U \Sigma V^*.$$

**Prueba**: Basada en Cornelious Lanczos.  Asumamos $A \in \mathbb{C}^{m \times n}$. Construyamos

$$ S = \begin{pmatrix} 0  & | & A  \\ -- & -- &  -- \\ A^* & | & 0 \end{pmatrix}  .$$


La matriz $S \in \mathbb{C}^{(m+n) \times (m + n)}$. Es autoadjunta
$S = S^*$,  por el teorema espectral de matrices auto-adjuntas
existen $w_i$ autovectores y $\lambda_i$ autovalores de $S$ tales que

$S w_i = \lambda_i w_i   \tag{11} $$,

Construimos el vector

$$w_i = \begin{pmatrix} u_i \\ v_i  \end{pmatrix}   .$$
con $u_i \in \mathbb{C}^{m}$ y $v_i \in \mathbb{C}^n$ ademas de
(11)

\begin{eqnarray}
A^* u_i = \lambda_i v_i \\
A v_i = \lambda_i u_i
\end{eqnarray}










Multipliquemos estas dos ultimas igualdades por $A$ la primera y por $A^*$ la segunda

\begin{eqnarray}
A A^* u_i = \lambda_i A v_i = \lambda_i (\lambda_i u_i) = \lambda_i^2 u_i \\
A^* A v_i = \lambda_i A^* u_i = \lambda_i \lambda_i v_i = \lambda_i^2 v_i
\end{eqnarray}
Es decir que $\lambda_i^2$ son autovalres de las matrices $A A^*$ y
$A^* A$

Como las matrices $A A^*$ y $A^* A$ son autoadjuntas los autovectores son ortogonales. Es decir los $u_i$ y los $v_i$.


Definamos dos matrices $U_r, V_r$ (de columnas de $u_i, v_i$  donde $r$ es el numero de autovalores que no son 0. De la misma forma definamos $U_0, V_0$ , matrices de columnas de $u_i , v_i$ pero para aqueyos autovalres que son 0.


Tenemos

\begin{eqnarray}
A^* U_r &=& V_r \Lambda_r \quad , \quad A^* U_0 = 0 \\
A V_r &=& U_r \Lambda_r \quad , \quad A V_0 = 0
\end{eqnarray}

Podemos construir lo siguiente

$$\Sigma = \begin{pmatrix} \Lambda_r & | & 0 \\  -- & -- & -- \\ 0 & | & 0  \end{pmatrix} $$

Juatando todo lo de arriba

$$ AV = U \Sigma .$$
Expliquemos esto de una mejor forma

$$AV = A [ V_r | V_0 ] = [A V_r | A V_0] = [U_r \Lambda_r | 0 ] = [U_r \Lambda_r + U_0 \times 0  | U_r \times 0 + U_0 \times 0] = U \Sigma.$$





De $AV = U \Sigma$ y como $U, V$ son ortogonales, multipliquemos a derecha por $V^*$


$$A = U \Sigma V^* .$$

Esto era lo que tocaba probar.
