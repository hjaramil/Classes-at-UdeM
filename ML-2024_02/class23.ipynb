# Bias vs. Variance (sesgo vs varianza)
Vamos a definir los terminos de forma "suelta" (sin rigor) y luego los definimos matematicamente (estadisticamente).

* **Bias**: Son errores que aparecen en ML debido a que los datos no se ajustan (en promedio) a la solucion. Las predicciones, en promedio, estan lejos del valor real. Esta relacionada con el "underfitting" (subajuste)

* **Variance**: (varianza). Es el error que ocurre cuando los datos se quieren ajustar mucho al modelo. Posiblemente siguiendo el ruido. Estadisticamente la varianza es que tan dispersos (regados, oscilan mucho) estan los datos. Esta relacionada con el "overfitting" (sobre ajuste).

* **Ruido**: En la practica **todos** los datos tiene ruido.

Vamos a formular el tradeoff (compromiso) entre el "bias" y el "variance".

Comecemos con funcion de costo de regresion lineal.

$$J(\theta)  = \frac{1}{2m} \left [ \sum_{i=1}^m ( h(\theta, x^{(i)})- y^{(i)})^2 + \lambda \sum_{j=1}^n \theta_i^2 \right ]$$

Vamos a ver que esta funcion de costo es un MSE + regularizacion.
MSE: Mean-Squared-Error, salvo un factor de $1/2$.

Consideremos un modelo con datos $x_1, x_2, \cdots, x_m$,
y etiquetas $y_1, y_2, \cdots, y_m$ asociadas con los puntos $x_i$. A cada $x_i$ le corresonde una etiqueta $y_i$, de forma
que podemos formula el problema como

$$y = f(x) + \epsilon $$
$\epsilon$ es ruido con media 0 y desviacion estandard $\sigma$.

La hipoteis es $h(\theta, x)$ no siempre se ajusta a las etiquetas $y_i$. En esta parte del curso, voy a escribir la hipotesis como $h(x)$ (no es importante $\theta$ para este analisis).  El
[MSE](https://en.wikipedia.org/wiki/Mean_squared_error)
como lo dice el nombre es la esperanza de $(y-h(x))^2$ al
es decir:

$$\mathbb{E}[ ( y - h(x))^2] = \sum_{i=1}^m ( y_i - h(x_i))^2 p(x_i) $$
donde $p(x)$ es la probabilidad de los datos $x_i$. Si la distribucion de los datos es uniforme $p(x)=1/m$
y

$$\mathbb{E}[ ( y - h(x))^2] = \frac{1}{m} \sum_{i=1}^m ( y_i - h(x_i))^2  $

Vamos a definir mas formalmente Bias, Variance
Por el momento asumimos que $\epsilon = 0$.

$$\text{Bias}[ h(x)] = \mathbb{E}[ h(x)] - \mathbb{E}[f(x)] $$
$$\text{Var}[h(x)] = \mathbb{E}[h^2(x)] - \mathbb{E}^2[h(x)] $$

Vamos a buscar una relacion entre estas cantidades tan importantes.

Estudiemos las siguientes ecuaciones

* MSE:
$$\mathbb{E}[ (y - h(x))^2] = \mathbb{E}[ y^2 - 2 y h(x) + h^2(x) ]  = y^2 - 2 y \mathbb{E}[ h(x)] + \mathbb{E}[ h^2(x)] \tag{1} $$

* Bias$^2$
\begin{eqnarray}
\text{Bias}^2[ h(x) ] &=& ( \mathbb{E}[h(x)]) - \mathbb{E}[f(x)])^2 \\
&=& \mathbb{E}^2[ h(x)] - 2 \mathbb{E}[h(x)] \mathbb{E}[f(x)] + \mathbb{E}^2 [ f(x) ] \\
&=& \mathbb{E}^2 [ h(x) ] - 2 y \mathbb{E}[h(x)] + y^2 \\
\end{eqnarray}

* Var$[h(x)]$
$$\text{Var}[h(x)] = \mathbb{E}[h^2(x)] - \mathbb{E}^2[h(x)] $$


Recordemos el MSE de la ecuacion
Para poder amarrar MSE, Bias, Variance sumamos y restamos
un termino como sigue:

$$\mathbb{E}[ (y - h(x))^2] = y^2 - 2 y \mathbb{E}[ h(x)] + \mathbb{E}^2[ h(x) ]  +  \mathbb{E}[ h^2(x)] - \mathbb{E}^2[h(x)]= \text{Bias}^2[h(x)] + \text{Var}[h(x)] $$



Encontramos que
$$\text{MSE}^2 = \text{Bias}^2 [ h(x) ] + \text{Var} [ h(x)] $$

Este resultado muestra que el error (mean squared, que es la funcion de costo de la regresion lineal) es la suma de dos errores: el Bias$^2$ y la varianza.

## Learning Curves (curvas de aprendizaje).

Son diagramas para evaluar el tipo de ajuste/subajuste/sobreajuste.


Antes de mostrar los diagramas recordemos estos conceptos:

* High bias (sesgo alto): Es lo que se ve en la figura de
la izquierda. Underfitting.

* High Variance: El grafico de la derecha. Se separan mucho los datos de entrenamiento con los de validacion. Hay varianza alta. No se acercan a asintota. Overfitting.


image.png

Como curar el sesgo y la varianza.

* Para curar problemas con varianza alta:
    * Mas datos.
    * Reducir el numero de parameros.  Por ejemplo en ANN reducir el numero de nodos por capa o reducir el numero de capas. En regresion polinomica, reducir el grado del polinomio.
    * Incrementar el valor de  $\lambda$.
* Para curar problemas de sesgo (bias):
    * Agregar hiperparametros. Mas capas en una red neuronal, mas nodos por capa, polinomios de grado mas alto.
    * Reducir el tamanho de $\lambda$.




# SVM: Support Vector Machines (Maquinas de soporte vectorial)
## Historia:
Este tema lo desarrollo Vladimir Vapnic (Ruso) por alla en los 60s. Fue trabajo de tesis. Vladimir emigro a los Estados Unidos por alla en 1992. El tema era desconocido. El sometio 3 articulos
a la revista:

[Conference on Neural Information Processing Systems](https://en.wikipedia.org/wiki/Conference_on_Neural_Information_Processing_Systems)

pero los 3 fueron rechazados. El trabajaba en los laboratorios Bell y aposto con un colega a que el metodo de el (SVM) superaba
a los metodos tradicionales para predecir digitos, como MNIST.
Vapnic gano la apuesta. La implementacion original del metodo no era eficiente y aparecio Isabel Guyon (introdujo el llamado
"kernel trick" , la semana entrante lo explico) y el metodo se popularizo.

Cual es la idea que tenia Vapnik sobre lo que es SVM.

En regresion lineal se quiere hallar una recta que se ajuste a una nube de puntos. La funcion de costo para este problema es

$$J(a,b) = \frac{1}{2m} \sum_{i=1}^m [ y_i - (a x_i + b)]^2 $$
minimizar esta funcion de costo.

En SVM se quiere **maximizar** el "gap" entre los puntos de un lado y los puntos del otro.

La funcion de costo para este problema

$$J(a,b) = \frac{1}{2m} \sum_{i=1}^m d^2[ (x_i, y_i), L(a,b)] $$
