Un objeto cae y una plataforma (por debajo) se mueve hacia arriba.
Encontrar el punto de encuentro del objeto y la plataforma asumiento
que sus velocidades son iguales (pero con signo contrario).

Matematicalmente

1. Objeto
$$ h_0(t) = h_0 - v_0 t - \frac12 g t^2 . \tag{1}$$
Salvedad: El subindice $_O$ es para objeto, no para "inicial"

2. La plataforma
Asumamos que la plataforma comienza a desplazarse desde el piso

$$h_p(t) = v_p t  + \frac12 q t^2 . \tag{2}$$

Tabla de simbolos

| variable | significado |
|-----------|------------|
| $t$ | tiempo |
| $h_0$ | altura inicial del objeto |
| $v_0$ | velocidad inicial del objeto |
| $v_p$ | velocidad de la plataforma|
| $g$ | constante de gravedad |
| $q$ | aceleracion de la plataforma |



```python
from sympy import Rational, symbols
t, h_0, v_0, g, v_p, q = symbols('t, h_0, v_0, g, v_p q', real=True, positive=True)
h0t = h_0 - v_0*t - Rational(1,2)*g*t**2 # ecuacion 1

# la velocidad es la derivada de la posicion
# derivadas podemos tomar en SymPy, pero esto viene mas adelante en esta clase o la proxima
dh0dt = -g*t + - v_0  # derivada a mano de la ecucaion (1)


# de igual forma la ecuacion (2)
hpt  = v_p*t + Rational(1,2)*q*t**2

# y sacando la derivada a mano
dhdpt =v_p + q*t


```

```python
# probamos lo que hemos hecho
display(h0t)
display(dh0dt)
display(hpt)
display(dhdpt)
```

```python
# Que ecuaciones necesitamos?
# se necesita que la particula encuentre la plataforma h0t=hpt
# que las velocidades sean iguales ( y de signo contrario) dh0dt=-dhpdt
eq1 = h0t-hpt  # las ecuaciones se igualan a 0
eq2 = dh0dt + dhdpt  # note que cambiamos el signo del segundo termino por ser direcciones contrarias

display(eq1)
display(eq2)
```

```python
# resolvemos las ecuaciones
# como son cuadraticas, devuelven dos valores
from sympy import solve
t_collide, v_intfall = solve( [eq1, eq2], [t, v_0])[0]
```

```python
slnes = solve( [eq1, eq2], [t, v_0])
```

```python
len(slnes)
```

```python
slnes[0]
```

```python
slnes[1]
```

```python
display(t_collide)
print("\n")
display(v_intfall)
```

```python
# que velocidad tienen los objetos cuando se encuentran
vh = dh0dt.subs( [(t, t_collide), (v_0, v_intfall)])
vh
```

```python
# simplificar
from sympy import simplify
simplify(vh)
```

```python
# todo en uno
vh = dh0dt.subs( [(t, t_collide), (v_0, v_intfall)]).simplify()
vh
```

Vamos a probar la siguiente conjetura


$$ \log_{10}x = \frac{\log x}{\log 10}. $$


\begin{eqnarray}
\log_{10} x = y &⟺& 10^y = x \\
 &⟺& \log( 10^y) = \log x \\
 &⟺&  y \log 10 = \log x \\
 &⟺&   y = \frac{\log x }{\log 10}
\end{eqnarray}

Lo que la celda de arriba dice

$$ \mathrm{e}^{\log_{10} x} ?.$$

Veamos

$$\mathrm{e}^{\frac{\log x}{\log 10}} = ( \mathrm{e}^{\log x})^{1/\log 10} =  x^{1/\log 10}.$$

### Identidades trigonometricas
Veamos algunas identidades de texto:



\begin{eqnarray}
\frac{1}{1 + \cot^2 \theta} &=& \sin^2 \theta \\
\frac{1}{1 + \tan^2 \theta} &=& \cos^2 \theta \\
\sin(\alpha \pm \beta) &=& \sin \alpha \cos \beta \pm \sin \beta \cos \alpha \\
\cos(\alpha \pm \beta) &=& \cos \alpha \cos \beta \mp \sin \beta \sin \alpha

Probamos estas identidades com `SimPy`.

## Limites
Ejemplo simple, limite por reemplazo.

Observe el siguiente procedimiento

$$\lim_{x \to 0} \frac{3 x^2 + 1}{x^2 - 1} = \frac{1}{-1}=-1 .$$

```python
from sympy import limit, symbols
x = symbols('x')
f = (3*x**2 + 1)/(x**2 - 1)

display(f)
limit(f, x, 0)
```

### hay cosas que no se pueden evaluar directamente
Por ejemplo

$$g(x) = \frac{x^2 + 4x}{x^2 - 2x} .$$
Si evalua obtiene $0/0$.

Hay dos formas de resolver esto

* Factorizando y cancelando

$$\lim_{x \to 0} g(x) = \lim_{x \to 0} \frac{x^2 + 4x}{x^2 - 2x} = \lim_{x \to 0} \frac{x(x+ 4 )}{x(x-2)} = \lim_{x \to 0} \frac{x+4}{x-2} = \frac{4}{-2}=-2.$$

* Regla de L'Hopital. Se saca la derivada arriba y abajo y se intenta de nuevo



$$\lim_{x \to 0} g(x) = \lim_{x \to 0} \frac{x^2 + 4x}{x^2 - 2x}= \lim_{x \to 0} \frac{2x + 4}{2x - 2} = \frac{4}{-2}=-2 .$$


Tambien sabemos ue

$$\lim_{x \to 0 } \frac{\sin x}{x} = 1 .$$
Esto lo verificamos en `SymPy`



```python
g = (x**2 + 4*x)/(x**2  - 2*x)
g.subs(x,0)
```

```python
limit(g, x, 0)
```
