## Integracion de un campo escalar a lo largo de una superficie.
### Teoria:
Sea unn campo vectorial $f=f(x,y,z)$. La integral del campo a lo largo de una superficie parametrizada por $r(u,v)=(x(u,v), y(u,v), z(u,v))$ esta dada por


$$ \int_S f dS = \int_S f(r) \left  \| \frac{\partial r}{\partial u}
\times \frac{\partial r}{\partial v} \right \| du dv .$$

donde

$$ dS  = \left \|  \frac{dr}{\partial u} \times \frac{dr}{\partial v}  \right \|  du d v.$$ es el elemento de area.

```python
from sympy import symbols, Function, Matrix,diff
import sympy as sp
u, v = symbols('u, v', pos=True)
x,y,z,f = symbols('x y z f', cls=Function, real=True)
x = x(u,v)
y = y(u,v)
z = z(u,v)
f = f(x,y,z)
r = Matrix( [x,y,z])
f
```

```python
integrand = f*sp.sqrt( diff(r,u).cross(diff(r,v)).dot(diff(r,u).cross(diff(r,v))))
integrand
```

$$ 1 + x^2 + y^2 = 1 + u^2 \cos^2 v + u^2 \sin^2 v = 1 + u^2 .$$

```python
integrand1 = integrand.subs( {
    f: sp.sqrt(1 + x**2 + y**2),
    x: u*sp.cos(v),
    y: u*sp.sin(v),
    z: v})

integrand1
```

```python
from sympy import integrate
integrand2 = integrand1.doit().simplify()
integrand2
```

```python
# evalumemos la integral
sln = integrate( integrand2, (u,0,2), (v, 0, sp.pi))
sln
```

Resolvamos el plano de forma **explicita**.

$$ z = \frac{6 - 2 u - v }{3} .$$
Finalmente necesittamos los limites de integracion. El primer octante implica que

* para $z=0$
$$ 2x + y = 6$$,
para $y=0$, $x=3$, y para $x=0$ , $y=6$. Encontramos que
$x \in [0, 3]$, $y \in [0,6]$.

```python
from sympy import solve
x,y,z= symbols('x y z ', cls=Function, real=True)
u,v=symbols('u v', pos=True)
F1,F2,F3= symbols('F1 F2 F3 ', cls=Function, real=True)

x = x(u,v)
y = y(u,v)
z = z(u,v)
F1 =  F1(x,y,z)
F2 =  F2(x,y,z)
F3 =  F3(x,y,z)

r = Matrix([x,y,z])
F = Matrix( [F1, F2, F3])
display(F)
eq1 = 2*x + y + 3*z - 6 # igualada a 0
slns = solve(eq1, z)
display(slns)
slns[0] # es una sola
zval = slns[0].subs(
    {x:u, y:v}
)
zval
```

```python
integrand = F.dot( diff(r,u).cross(diff(r,v)))
integrand
```

```python
u, v = symbols('u v', pos=True)
x,y,z, F1, F2, F3= symbols('x y z F1 F2 F3', cls=Function, real=True)
x = x(u,v)
y = y(u,v)
z = z(u,v)
F1 = F1(x,y,z)
F2 = F2(x,y,z)
F3 = F3(x,y,z)
integrand1 = integrand.subs({
    F1:y ,
    F2: x,
    F3: z
})
integrand1
```

```python
integrand2 = integrand1.subs(
    {
        x:u,
        y:v,
        z:zval
    }
).doit()
integrand2
```

```python
sln = integrate( integrand2, (u, 0, 3), (v, 0, 6))
sln
```

# Teoremas importantes del calculo en varias variables.
## Teorema de Green.
Sea $C$ una curva cerrada orientada positivamente. Es decir, cuando usted camina sobre la curva la region esta a su izquierda.
Sean $P,Q$ dos campos vectoriales en $\mathbb{R}^2$ y $D$ la region acotada por el camino. Entonces


$$\oint P dx + Q dy = \iint_\limits D \left ( \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} \right ) d A.$$

Ejemplo: Evalue la integral $\oint_C (x^5 + 3y)dx + (2x - \mathrm{e}^{y^3})dy $, donde $C$ es el circulo con entro en $(1,2)$ y radio 2. Es decir $(x-1)^2 + (y-2)^2 = 4$.

Hacemos esto a mano y luego con `SymPy`

El calculo de la izquierda es imposible analiticamente. Por la funcion $\mathrm{e}^{y^3}$

```python
integrate(sp.exp(x**3), (x,0,1))
```

Como vemos la integral de la izquierda tiene problemas, por que esta en funcion de funciones especiales Gamma $\Gamma$, gamma  $\gamma$.

Evaluamos entonces el lado derecho.

Dado que $P=x^3 + 3y$, $Q=2x - \mathrm{e}^{y^3} $
entonces

$$\frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} = 2 - 3 = -1 .$$
El integrando es $-1$ sobre al area.

$$-\iint\limits D dA = -A $$
El area es aqueya de un circulo de radio 2, es decir $A=\pi r^2 = 4 \pi$. El resultado es

$$ \oint_C P dx + Q dy = - 4 \pi .$$

```python
t = symbols('t', pos=True)
x,y,P,Q = symbols('x y P  Q', cls=Function, real=True)

# integral a la izquierda, un circulo parametrizado en t
# circulo con centro (1,2) y radio
x = 2*sp.cos(t) + 1
y = 2*sp.sin(t) + 2

P = x**5 + 3*y
Q = 2*x - sp.exp(y**3)

integrand = P*diff(x,t) + Q* diff(y,t)

int1 = integrate( integrand, (t, 0, 2*sp.pi))
int1
```

```python
# no lo hizo
int2 = int1.simplify()
int2
```

```python
leftHS = int2.evalf() # resuelva la integral de forma numerica
leftHS
```

```python
# forcemos la respuesta en terminos de pi
from sympy import nsimplify
nsimplify( leftHS, [sp.pi], tolerance=0.01)
```

```python
# lado derecho
P, Q = symbols('P Q', cls=Function, real=True)
x,y = symbols('x y ', real=True)

P = x**5 + 3*y
Q = 2*x - sp.exp(y**3)
integrand = diff(Q,x)-diff(P,y)
integrand
```

```python
# integramos sobre el circulo
# yo puedo mover circulo al origen, por que el integrando es constante.
r, theta = symbols('r theta', pos=True)
intVal = integrate(integrand, (r, 0, 2), (theta, 0, 2*sp.pi))
intVal
```
