## Factorizacion sobre los complejos.

Recuerde
$$x^2 + 1  = (x + \mathrm{i})(x - \mathrm{i}).$$
donde $\mathrm{i}=\sqrt{-1}$.
El teorema fundamental del algebra nos dice que todo polinomio es completamente factorizable sobre los complejos.


````python
# factoricemos x^2 + 1 sobre los complejos
from sympy import I
factor(x**2 + 1, extension=[I])
````

````python
h2 = x**3 - x + 8
factor(x**3-x+8, extension=[I])
````

### Actividad 1:
Investigue por que no funciono la factorizacion anterior
y como se arregla este problema.

Otra forma de factorizar es hallando las raices.

Por ejemplo:
$$ x^2 -x - 6.$$

$$ x= \frac{1 \pm \sqrt{1 + 24 }}{2} .$$
Claramente son

$$x_1 = 3, -2.$$

La factorizacion es $(x-x_1)(x- x_2)$.
Es decir
$$(x-3)(x+2).$$

````python
# hallamos las raices y factorizamos
from sympy import roots
r = roots(h2)
#display(r)
r
````

````python
# factoricemos
hfactor = h2.factor(extension= r)
hfactor
````

Lo que estamos viendo es $(x-r_1)(x-r_2)(x-r_3)$ donde $r_i$, $i=1,2,3$ son
las raices de $h2$.

### simplicar

````python
from sympy import simplify
# simplifiquemos las raices
rsimp = simplify(r)
rsimp
````

````python
# podemos simplificar hfactor ?
simplify(hfactor)
````

### Actividad 2:
La factorizacion de $h2$ esta dada por


$$(x-r_1)(x-r_2)(x -r_3) . \tag{1} $$
con $r_i$, $i=1,2,3$ las raices de **hfactor** ($x^3 - x + 8)$.

Genere la factorizacion de $h2$ usando la formula (1) con las raices $r_i$, $i=1,2,3$ **simplificadas**.

Al final muestre que $$(x-r_1)(x-r_2)(x-r_3)=x^3 - x + 8.$$

## Collect.
Considere la siguiente expresion

$$xy + xz + x^2 y z + yz .$$
Por alguna razon queremos formar un polinmimo en $x$. Entonces


$$p(x) = yz + x(y+z) + x^2 yz  .$$

````python
from sympy import symbols, collect
x,y,z = symbols('x y z ')
expr = x*y + x*z + x**2* y*z + y*z

coll = collect(expr, x)
coll
````

````python
# hallemos coeficientes, Otra forma de "parsing"
c= coll.coeff(x, 1)
c
````

````python
z1 = coll.coeff(x,2)
z1
````
