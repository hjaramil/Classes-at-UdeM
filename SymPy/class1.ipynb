### Continued Fraciones

#Construimos fracciones continuadas de la forma

<font size="12">
\begin{eqnarray}
a_0 + \frac{x}{  a_1 + \frac{x}{ a_2 + \frac{x}{a_3 + \frac{x}{a_4 + \frac{x}{a_5 + \cdots }}}}}
\end{eqnarray}
</font>

````python
from sympy  import symbols
a = symbols('a0:6')
a
````

````python
x= symbols('x')
````

````python
def continued_fraction(depth):
    frac = a[0]
    for d in range(depth):
        frac = frac.replace( a[d], a[d] + x/a[d+1])
    return frac
````

````python
cf = continued_fraction(5)
cf
````

Si quiere converetir la anterior expresion a LaTeX lo puede hacer

````python
sm.latex(cf)
````
