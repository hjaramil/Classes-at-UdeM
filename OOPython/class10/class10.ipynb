###  Asociacion:
Es otra forma de relacionar objetos.

Aca hay una asociacion entre un vuelo  y  un vaion. Explico los simbolos
* 0..1: En este lado hay  o ninguno o 1 maximo.
*  0...n :  En este caso hay ninguno o hasta n maximo
* 1: obligatorio que hay  solo  1
*  1..n. o $1..^*$  Es obligatorio que hay 1 o  mas, hasta  n.


Dos tipos de asociacion:

* Agregacion (aggretation).  Este es "tiene un"  .  "has a". Tambien se llama  ***containment** (contenido) en este caso,  una canasta tiene naranjas.  Selection_001.png

* Composition: "is part of" (es parte de).  Ejemplo del cuerpo y el corazon. Sin cuerpo no hay corazon, el carro y  la silla,  sin carro no hay silla, o no hay frenos.  Selection_002.png

### Que diablos hace el self?
La sintaxis para definir una clase es:
```class NameofClass: ```


````python
class Square:
    side =2 

    def sqr(this):
        print("address of the objet", id(this))
        print()
        return this.side**2  # find the area

mySqr = Square()
mySqr.sqr()
print(id(mySqr))
````

````python
class Square:
    side =2 

    def sqr(self):
        print("address of the objet", id(self))
        print()
        return self.side**2  # find the area

mySqr = Square()
mySqr.sqr()
print(id(mySqr))
````
