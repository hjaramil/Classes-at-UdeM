# Inhertitance (Herencia)
## Multiple inheritance (Herencia Multiple)
###  MRO: Multi-resolution order.  Orden de multi-resolucion.

````python
class A():
    def callMe(self):
        print("A")
        return

class B(A):
    def callMe(self):
        super().callMe()
        print("B")
        return\

class C(A):
    def callMe(self):
        super().callMe()
        print("C")
        return

class D(B,C):
    def callMe(self):
        super().callMe()
        print("D")
        return
        
````

````python
myD = D()
myD.callMe()
````

````python
help(D)
````

````python
class A():
    def callMe(self):
        print("A")
        return

class B(A):
    def callMe(self):
        super().callMe()
        print("B")
        return\

class C(A):
    def callMe(self):
        super().callMe()
        print("C")
        return

class D(C,B):
    def callMe(self):
        super().callMe()
        print("D")
        return
        
````

## Polimorfismo (Polymorphism)

Una funcion o un operador pueden tomar distintas formas de acuerdo al contexto ("duck typing").



````python
class Animal:

    def __init__(self, name):
        self.name = name
        return

    def move(self):
        print("I %s am moving"%self.name)
        return

class Fish(Animal):

    def move(self):
        print("I am a %s, and I am swiming "%self.name)
        return

class Snake(Animal):
    def move(self):
        print("I am a %s and I crawl"%self.name)
        return

myFish = Fish("fish")
myFish.__dict__
````

````python
myFish.move()
````

````python
mySnake = Snake("snake")
mySnake.move()
````

El ejemplo anterior sigue siendo **overriding** (sobre escribir). De nuevo, overriding esta en la interseccion entre herencia y polimorfismo.

## Overloading
Este concepto es parte de polimorfismo pero no de herencia.
En overloading el nombre es el mismo pero el "signature" es distinto:

"Signature:": los argumentos y los tipos de argumentos, y el retorno.

signature=firma.


Python no maneja "overloading" (sobrecarga).
Veamos varios ejemplos:

En alguna aplicacion necesitamos sumar dos terminos o tres de acuerdo al contexto, pensemos en el programa sum

````python
class Sum:

    def __init__(self):
        return

    def sum(self, x, y ):
        return x+y

    def sum(self, x, y, z ):
        return x + y + z

s = Sum()
print(s.sum(2,3,4))
````

````python
print(s.sum(2,3))
````

Esto si se puede hacer en python pero tiene un truco. Mas tarde (en 4 clases o mas) lo muestro

Hay otra forma de hacer sobrecarga, pero con "if"
Hagamos "overload" del ```__init__()```

````python
class Student():

    # first __init__
    def __init__(self):
        return
    
    def __init__(self, name, score):
        self.name = name

        if (score<0):
            print("invalid score. It should be a positive number")
        else:
            self.score = score
        return

st = Student()
````

````python
st = Student("Oscar", 4)
st.__dict__
````

Una forma de arreglar el problema de polimorfisco es usano "if". No es elegante

````python
class Student2():

    def __init__(self, *args): # argumentos variables
        if len(args) ==0 :
            self.name = "Jack"
            self.score = 0
        if len(args) == 2:
            self.name = args[0]
            if args[1] < 0 :
                print("scores should be positive")
            else:
                self.score=args[1]
        return

st = Student2()
st.__dict__
````

````python
st2 = Student2("George", 4)
st2.__dict__
````

````python
class Complex:

    def __init__(self, x, y):
        self.x = x
        self.y = y
        return

    def __add__(self, a):
        x = self.x + a.x
        y = self.y + a.y
        return (x,y)

    def __mul__(self, a):
        x = self.x*a.x - self.y*a.y
        y = self.y*a.x + self.x*a.y
        return (x,y)
````

````python
a = Complex(5,2)
b = Complex(-1,3)
a.__dict__
````

````python
print(vars(b))
print("the sum of a+b is", a+b)
print("the product of a*b is ",a*b)
````
