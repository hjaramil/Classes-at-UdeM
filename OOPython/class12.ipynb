# Encapsulacion

 Se trata de "encapsular" los datos y los comportamientos en un solo bloque (clas). Tiene muchas ventajas:
 

 * proteger los datos nuestros. Seguridad contra nosotros mismos y contra otros. Por privacidad, por confidencialidad, propiedad intelectual.

 * proteccion contra otros. En mi carro yo puedo ver la temperatura del motor en un panel. Que diablos le importa a los conductores de al lado o detras saber esta informacion. No solo no es relevante sino distractora. 

 * modularidad: En modulos es mas facil trabajar. Son rompecabezas , legos. El ejemplo de la estacion espacial. 

 * Reuso, reciclable. 
 * Esconder complejidad. El capo del carro nos tapa la complejidad.

# Como ver los datos "privados"
Una tecnica es con "getters()" y "setters()" (mutators).



````python
class Car:
    def __init__(self, brand, model, year, engineTemperature):
        self.brand = brand
        self.model = model
        self.year = year
        self.__engineTemperature = engineTemperature
        return

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1

    def getTemp(self):
        return self.__engineTemperature

    def setTemp(self, newTemp):
        self.__engineTemperature = newTemp
        return

test = Car("Toyota", "Corola", 2016, 79)
test.temperaturePlusOne()

print(test.brand)
print(test.model)
print(test.year)

print(test.getTemp()) # esta instruccion debe funcionar

````

````python
# la podemos cambiar.
test.setTemp(90)
test.getTemp()
````

## overload the ```__init__()``` method

````python
class Student():

    # first init
    def __init__(self):  # no parameters
        self.name = "jack"
        self.score = 0
        return


    # second init
    def __init__(self, name, score):
        self.name = name
        if (score<0):
            print("invalid score. Shoule be postive")
        else:
            self.score = score

        return

s1 = Student()
````
