Como representamos los numeros con simbolos y cifras.

Ejemplo: en base 10,  $10.235=1 \times 10^1 + 0 \times 10^0 + 2 \times 10^{-1} + 3 \times 10^{-2} + 5 \times 10^{-3} $ 

En general: para base $p$

$$a_1 a_2 \cdots a_n . b_1 b_2 \cdots b_m = a_n \times p^0 + a_{n-1} p^1 \cdots a_1 p^n +  b_1 \times 10^{-1} + b_2 \times 10^{-2} + \cdots b_m \times 10^{-m}  $$


Con solo dos simbolos que posibilidades tenemos: $a=0, b=1$, tres cifras.

\begin{eqnarray}
aaa=000 = 0\\
aab=001 = 1\\
aba = 010 = 2 \\
abb = 011 = 3\\
baa = 100 = 4\\
bab = 101 = 5\\
bba = 110  = 6\\
bbb = 111 = 7
\end{eqnarray}

Un poco de combinatoria.

Con base $n$ ($n$ simbolos) y $m$ cifras tenemos un total de combinaciones
de $n^m$ numeros. Estos numeros son $0,1, \cdots n^{m}-1$.
