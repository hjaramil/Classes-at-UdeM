#### Slicing.
Sintaxis:

`[start: end: incremen]`, el incremento puede ser negativo.

* Si `increment` no esta dado se asume 1
* Si el `start` no esta dado se asume el primero elemento. Indice 0
* Si el `end` no esta dado se asume el ulitmo elemento $n-1$. Donde $n$ es la longitud de la lista.
* Para accesar el indice `i` se llama con `a[i]` donde `a` es el nombre de la lista

```python
myList[2:6:2]
```

Algunas funciones importantes de estas son

* `append` : para agregar al final
* `insert` : agregar luego del indice `i`.
* `remove` : remover el elemento del indice `i`
* `pop`  : una forma rapida de remover el ultimo elemento
* `sort` : ordena la lista
* `reverse` : reversa el orden de una lista


```python
myList = list( [3,2])  # [3,2]
print("mylist is", myList)
myList.append("hola")
print("appending 'hola' a myList",myList)

myList.insert(10, 'que tal')
print("inserting 'que tal' after element 10 in myList",myList)

myList.remove('hola')
print("luego de remover 'hola' de la lista", myList)

myList = [3,5,-1,5, 0]  # se reescribe. Las listas son mutables. 
print("sobreescrimos la lista. Ahora es:", myList)

# sort
myList.sort()
print("luego de ordenarla", myList)
myList = ["coco", "abc", "hello"]
myList.sort()
print("luego de ordernarla", myList)

myList.reverse()
print("al reves", myList)





```

### List comprehension
La sintaxis es:

`[expression for iterm in iterable if condition==True]`


* La `expression` es funcion del `item` (es un indice0
* El `item` es el indice, la variable independiente
* `iterable` es una lista, un rango, una tupla, o cualquier variable iterable
* `condition` es un filtro. Si `condition==True` ese elemento pasa, sino se queda.

```python
myList = [i**2 for i in range(10) if i%2]
```

```python
myList
```

```python
myList = [i**2 for i in range(10)]  # sin condicion
```

```python
myList
```

## mas de contenedores.
Como saber si un objeto es contenedor o no.
En el siguiente ejemplo vamos a verificar si la lista [1,2,3] es o no un contenedor

```python
import numpy as np
a = np.array([1,2,3])
memebersOfa = dir(a)

if memebersOfa.count( "__contains__"):
    print("a is a container")
```

```python
memebersOfa
```

```python
p = [1, 3, 5 ,1]
p.count(1)
```

```python
memebersOfa.count("__contains__")
```
