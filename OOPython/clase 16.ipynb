# Agregacion (Aggregation)
Las palabras claves son "has a". 
Ejemplo con segmento

```python
# Clase Point

class Point:
    def __init__(self , x, y):
        self.x = x
        self.y = y
        return

    def norm(self):
        return ((self.x)**2 + (self.y)**2 )**(0.5)


# defina la clase segment
class Segment:
    def __init__(self, x1, y1, point):
        self.x1 = x1
        self.y1 = y1

        # chequeo.
        if not isinstance(point, Point):
            print("la sintaxis es x,y, point en arguments")
            return

        self.point = point

        # mover el punto al origen
        point.x = point.x - self.x1
        point.y = point.y - self.y1

        return


    def lengthSegment(self):
        return self.point.norm()

point = Point(5,6)
seg = Segment(3,4, point) # agregacion

print(seg.lengthSegment())

   
```

```python
import matplotlib.pyplot as plt


class Segment:

    def __init__(self, p1, p2, color="blue"):
        self.p1 = p1
        self.p2 = p2
        self.color = color
        return

    def length(self):
        return sqrt( (self.p2.x - self.p1.x)**2 + (self.p2.y - self.p1.y)**2)


p1 = Point(3,4)   
p2 = Point(2,3)
s1 = Segment(p1,p2, "red")
print(s1.p2)



    
```

```python
# clase triangulo
class Triangle():

    def __init__(self, p1, p2, p3, color="blue"):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.color=color

        self.side1 = Segment(p1,p2)
        self.side2 = Segment(p2,p3)
        self.side3 = Segment(p3,p1)
        return

    def perimeter(self):
        # es la suma de las longitudes de los tres lados
        return self.side1.length() + self.side2.length() + self.side3.length()

    def lengthSides(self):
        print(f"side length between point {self.p1} and point {self.p2}, {self.side1.length()}")
        print(f"side length between point {self.p2} and point {self.p3}, {self.side2.length()}")
        print(f"side length between point {self.p3} and point {self.p1}, {self.side3.length()}")

        return

    def findArea(self):
        vect1 = self.p3 - self.p1
        vect2 = self.p2 - self.p1
        
        areaMatrix = array([vect1, vect2])
        return 0.5*abs( det( areaMatrix))





```
