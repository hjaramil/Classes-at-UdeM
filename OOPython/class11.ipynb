# Metodo ```__init__()```
En clases anteriores hablamos del **constructor** . El **constructor** es un metodo que, una vez se instancia la clase, asume todos los parametros de la misma. 

Ejemplo

````python
# dunder methods (double underscore)
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

# instanciamos la clase
myCar = Car("Toyota", "colola", 2015)
````

````python
myCar.__dict__
````

Que esta detras de ```__init__()``` ?
Esta otro metodo llamado ```__new__()```. Para algunos autores este es el constructor de Python.

````python
# creamos el objeto carro usando new
myCarNew = object.__new__(Car, "Nissan", "Sentra", 2017)
myCarNew.__init__("Nisan", "Sentra", 2017)
````

````python
myCarNew.__dict__
````

````python
# otro ejemplo donde combinamos ambos.
class Car:
    
    def __new__(cls, brand, model, year):
        print(f"Creando un nuevo objeto  {cls.__name__}  ")
        obj = object.__new__(cls)
        print(obj)
        return obj
        
    def __init__(self, brand, model, year):
        print("inicializando con __init__")
        self.brand = brand
        self.model = model
        self.year = year

# instanciamos la clase en un objeto
myCar = Car("Toyota", "Corola", 2018)
print(id(myCar))
````

````python
# ultimo ejemplo de classe
class Square:
    def __init__(self, side, color, center):
        self.side = side
        self.color = color
        self.center = center

    def area(self):
        return self.side**2

    def imprima(self):
        print(" la clase Square se instancia  \
        con centro en %s, lado %f, \
        con color %s"%(self.center, self.side, self.color))

mySqr = Square(2, "amarillo", (2,3))
mySqr.imprima()

mySqr.area()

        
````
