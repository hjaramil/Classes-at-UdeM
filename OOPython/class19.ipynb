## Empalme con la clase anterior.

````python
# simple program of simple inheritance
class Person:
    def __init__(self, firstName, lastName, birthDay):
        self.firstName = firstName
        self.lastName = lastName
        self.birthDay = birthDay
        
# creamos un objeto
myPerson = Person("Juan", "Lopez", "06/21/84")
vars(myPerson)
````

````python
class Student(Person):
    def pleaseSit(self):
        print("please take a sit")
        return

p1 = Person("Gloria", "Hoyos", "13/08/61")
s1 = Student("Maria", "Mejia", "20/10/76")
print(vars(p1))
print(vars(s1))
````

````python
s1.pleaseSit()
````

````python
p1.pleaseSit()
````

Veamos un caso donde el ```__init__()``` es necesario por que hay mas parametros en la clase hija.|

````python
class Student(Person):

    def __init__(self, firstName, lastName, birthDay, grade):
        super().__init__(firstName, lastName, birthDay)
        self.grade = grade
        return

myStudent = Student("Jorge", "Agudelo", "05/21/1990", 11)
myStudent.__dict__
````

el metodo ```super()``` no es exclusivo del ```__init__()```, se puede usar en conjunto con otros metodos. Se deja como ejercicio al estudiante crear una clase donde un metodo que no sea el ```__init__()``` use el metodo ```super()```.

## El uso de herencia para crear interfaces informales (informal interfaces).




````python
# informal interface
# much like C++, C, Java, it is very typed.
class Animal:
    def move(self, name:str, start:float, end:float, time:float) -> None:
        pass

    def oralSound(self, name:str, volume:float)->None:
        pass

    def jump(self, name:str, height:float) ->None:
        pass

class Perro(Animal):
    def __init__(self, nameMove, start, end, time, nameSound, volume):
        ''' Esto es para lo documentacion del programa '''
        self.nameMove = nameMove
        self.start = start  # in meters
        self.end = end  # in meters
        self.time = time # in seconds
        self.nameSound = nameSound
        self.volume = volume # in dB
        return 

    def move(self):
        distance = self.end - self.start
        averageSpeed= distance/self.time
        print("A dog %s at a speed of %f m/s \n"%(self.nameMove, averageSpeed))
        return

    def oralSound(self):
        name=self.nameSound
        print("A dog %s at a volume of %f db\n"%(name, self.volume))
        return

    def jump(self):
        raise NotImplementedError

myAnimal = Animal()  # va a haber problemas
print(dir(myAnimal))
````

````python
myPerro = Perro("runs", 10, 20, 2, "bark", 50)
myPerro.__dict__
````

````python
myPerro.move()
````

````python
myPerro.oralSound()
````

````python
myPerro.jump()
````

## Ejemplo de herencia simple en "paralelo"



````python
from statistics import mean, median, mode
class Group:
    weights = [0.3, 0.4, 0.4] # class variable

    def __init__(self, students):
        self.students = students
        return
        
class Students(Group):
    def __init__(self, hw, partial, final, name, identification):
        self.hw = hw
        self.partial = partial
        self.final = final
        self.name = name
        self.id = identification
        return

    def finalGrade(self):
        return self.hw*self.weights[0]+self.partial*self.weights[1]+self.final*self.weights[2]


class Grades:
    def __init__(self, myList):
        self.myList = myList
        return

    def minim(self):
        return min(self.myList)

    def maxim(self):
        return max(self.myList)

class Statistics(Grades):
    def meanValue(self):
        return mean(self.myList)

    def medianValue(self):
        return median(self.myList)

    def modeValue(self):
        return mode(self.myList)

juan = Students(3,3.4, 2.8, "Juan", "23453")
pedro = Students(2.1, 2.2, 4.5, "Pedro", "35244")

print(juan.__dict__)
print(pedro.__dict__)
````

````python
juan.finalGrade()
````

````python
pedro.finalGrade()
````

````python
myStudents = [juan, pedro]

myClass = []
for student in myStudents:
    myClass.append(Group(student))
````

````python
myClass[0].students.__dict__
````

````python
myGrades = []
for i in range(len(myClass)):
    myGrades.append(myClass[i].students.finalGrade())

myGrades
````

````python
cal = Grades(myGrades)
print("nota minima", cal.minim())
print("nota maxima", cal.maxim())
````

````python
cal.meanValue()
````

````python
calStat = Statistics(myGrades)
print("nota minima", calStat.minim())
print("nota maxima", calStat.maxim())

meanv = calStat.meanValue()
medianv = calStat.medianValue()
modev = calStat.modeValue()

print("mean value", meanv)
print("median value", medianv)
print("mode value", modev)
````
