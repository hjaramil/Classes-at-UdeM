Ejemplo de como borrar un metodo de un objeto. A veces es necesario.

````python
class Car:
    brand:"Toyota"
    model="Corolla"
    year = 2014

    def __init__(self, color="yellow"):
        self.color=color
        return

    def stop(self):
        print("please stop")
        return

        
````

````python
c = Car()
c.color
````

````python
setattr(c, "color", "blue")
c.color
````

````python
vars(c)
````

````python
delattr(c, "color")
vars(c)
````

````python
d = Car()
d.color
````

````python
vars(d)
````

````python
del(d)
````

````python
e=Car()
e.stop()
````

````python
del Car.stop
e.stop()
````

## Variables de instancia: Son las que se definen en ```__init__()``` .
