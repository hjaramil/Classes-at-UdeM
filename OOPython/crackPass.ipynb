```python
%%time
import itertools
from itertools import product

digits = string.digits
letters = string.ascii_letters

# lettersDigits = letters+digits

# products = product(letters, repeat=6)
products0 = product(digits,letters, letters,  letters, letters, letters)
products1 = product(letters, digits, letters,  letters, letters, letters)
products2 = product(letters, letters, digits,  letters, letters, letters)
products3 = product(letters, letters, letters,  digits, letters, letters)
products4 = product(letters, letters, letters,  letters, digits, letters)
products5 = product(letters, letters, letters,  letters, letters, digits)

products = itertools.chain(products0,products1, products2, products3,products4,products5)


while True:
    obj = list(next(products))
    # print(passwd, obj)
    if passwd == obj:
        print("found it!")
        print(obj)
        break
        
```
