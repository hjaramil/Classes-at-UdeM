Los objetos son la instanciacion de las clases. La clase es un mapa que se usa para generar objetos. Es como una patente. Los objetos se generan en la fabrica basados en un diseo que es la clase.  Las clases (y los objetos por ser entidades generadas por las clases) constan de atributos (propiedades y metodos (comportamientos, verbos). La encapsulacion es la construccion de una clase como un conjunto cerrado de  propiedades y comportamientos.

###  Interfaces: 
Las interfaces son los puentes de comunicacion entre los  objetos. Todos atributos publicos estan en la interface. Por ejemplo la relacion entre un televisor (TV) y un televidente se puede establecer a traves del control remoto.

La **abstraccion** consite en representar los atributos y metodos  de los objetos. Hay muchos niveles de abstraccion. Entre un carro y un mecanico, entre un carro y un conductor, entre un doctor  y el corazon, entre un estilista y el  cabello.  

El  **constructor**  (constructor). Es un metodo que  corre automaticamente  cuando se instancia un objeto. En el constructor se pueden definir los parametros por defecto y otros atributos que no necesitan ser definidos explicitamente luego.    De a misma forma existe el **destructor** que simplemente elimina (borra memoria) . Este no se usa en python por que python tiene "garbage collector". El constructor/destructor se usan mucho en Java, C++, C# etc. 

**accessors**: metodos de acceso. Es comun tener atributos ocultos al usuario. Se crean interfaces para accesar esos parametros. Hay dos clases tipicas de interfaces de acceso.

1. Getters: (obtenibles,  leibles).  Esos leen los atributos de una clase. 
2. Setters:  (modificables) . Los  atributos  se pueden cambiar  mediante  estos  metodos.  

Estos metodos de acceso son muy  usados en Java,  C++, etc,  pero  no  en python.  

## Relacion de herencia: Inheritance. 
El reino animal tiene muchos atributos y  comportamientos los cuales es conveniente heredar a todos los animales. Por ejemplo a los mamiferos. 
Es decir un mamifero, por el solo hecho de ser animal, tiene una cantidad de comportamientos que se heredan del reino animal. Igualmente un perro puede heredar todos los comportamientos de un mamifero, etc. 
Asi como el hijo hereda del padre tenemos la clase **padre** (parent) le hereda a la clase **hijo**  (child). Otra palabra para la clase padre es **superclase**  (superclass),  y  la para  la clase hijo **subclase**  (subclass). A la clase hijo tambien se le llama **derivada** (derived class).

La superclase tiene un nivel mas alto de abstraccion. 
Comparanto esto con la factorizacion algebraica

$$a(x+y+z+w) = ax + ay + az + aw  $$

**is a relationship** (es una, heredar). Mientras que  del padre al hijo se hereda, del hijo al padre se **generaliza**.  Esto llama a  la relacion llamada **generalizacion**  (generalization).
