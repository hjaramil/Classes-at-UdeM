````python
# atributos de clase (variables de clase)
class Car:
    brand="Toyota"
    model="Corolla"
    year="2014"

    # en mis notas de clase no tengo init, pero lo recomiento
    def __init__(self, color="yellow"):
        self.color=color
        return

myCar = Car()
vars(myCar)
````

````python
delattr(myCar, "color")
````

````python
myCar.color
````

````python
vars(myCar)
````

````python
c= Car()
vars(c)
````

````python
del(c.color)
````

# geometria

````python
# clase cilindro
class Cylinder:
    pi = 3.14159

    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        return 

    def volumeCylinder(self):
        return self.pi * self.radius**2*self.height

    def surfaceArea(self):
        return self.pi * 2* self.radius*self.height


        
````

````python
c = Cylinder(3, 4)
c.volumeCylinder()
````

````python
3.1415*3**2*4
````

````python
c.surfaceArea()
````

````python
delattr(Cylinder, "pi")
````
