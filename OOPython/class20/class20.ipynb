Podemos relacionar la estructura anterior a la siguiente estructura algebraica
$$ a[ x(\alpha_1 + \beta_1) + y(\alpha_2 + \beta_2) + z(\alpha_3 + \beta_3)] = a x \alpha_1 + a x \beta_1 + a y \alpha_2 + a y \beta_2 + a z \alpha_3 + a z \beta_3$$

Para calcular la operacion de la izquierda se necesitan 5 sumas y 4 multiplicaciones, para la de la derecha se necesitan 5 sumas y 12 multiplicaciones. El punto no es que se ahorra operaciones (aunque esto es importante) sino que piense en que un codigo se puede factorizar y cuanto se gana factorizandolo de esta forma, en numero de lineas de codigo, en numero de operaciones, en memoria, etc.

Dejamos como tarea al estudiante la implementacion de un metodo con la estructura mostrada en la grafica de arriba
