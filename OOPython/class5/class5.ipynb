Vamos a sumar los numeros de una lista.

```
sumlist([],0)
sumlist([H,T], N) :- sumlist(T, N1), N is N1 + H.
```

```
sumlist([1,3,5,4]).
N=13
```
* En la lista ``` [1,3,5,4]``` la cabeza es ``` H=1``` y la cola es ```T=[3,5,4]```. Entonces ``` N= N1 + H = N1 + 1```  (eq 1)

* Aca tenemos como cabeza ``` H=3``` y la cola ```T = [5,4]``` . La ecuacion ```N=N1+3```. (eq 2)

*  Aca tenemos como cabeza ``` H=5``` y la cola ```T = [4]``` . La ecuacion ```N=N1+5```. (eq 3)

* Aca tenemos como cabeza ``` H=4``` y la cola ```T = []``` . La ecuacion ```N=N1+4```. (eq 4)

* No hay cabeza, lo que teneos es []. Usamos la primera linea y ```N1=0```.

*  Nos devolvemos:
    * de la eq 4. N=0+4, redefinimos N1=N=4.
    * de la eq 3. N1 = 4+5=9
    * de la eq 2. N1 = 9+3 = 12
    * de la eq 1. N1 = 12 + 1 = 13.



### functional programming (programacion funcional). Es un paradigma.

* Funciones puras. Como en matematicas. Entran parametros y arroja un resultado.
* No tiene efectos laterales (side effects). Las variables son inmutables. 
* Recursion en vez de ciclos. 

Ventajas:
* Es mas facil de depurar, por que las variables son inmutables.
* Las funciones solo retornan los mismos valores para los mismos argumentos. Lo hace consistente
* Memory cache. Como las funciones solo  retornan el  mismo valor, para  los mismos argumentos,  estos  valores se pueden guardar en una "look  up table".
Luego se pueden leer de alli cuando  ls funcion se llama con los mismos  argumentos, esto acelera el  proceso. 

Desventajas:
* La memoria puede crecer mucho por varias razones.
    * Como las variables son inmutables, piense en una lista que crece mucho.
    Piense en una lista L que crece dinamicamente. Si no se puede reciclar el numbre, habrian muchas listas parecidas ocupando mucha memoria.
    * La recursion puede generar un stack muy grande y el stack ocupa memoria.


Ejemplos de dos paradigmas de programacion. Imperativo estructurado funcional.

````python
# de procedimientos
a=5
def addone(x):
    return x+1

def addonethreetimes(x):
    x = addone(x)
    print("x=",x)
    x = addone(x)
    print("x=",x)
    x = addone(x)
    print("x=",x)

    return x

print("a antes de la mutacion", a)
a = addonethreetimes(a)
print("a despues de la mutacion",a )
````

````python
# sin mutacion
# functional programming
def addonethreetimesfunc(a):
    b = addone(a)
    print("b=",b)
    b = addone(a)
    print("b=",b)
    b = addone(a)
    print("b=",b)
    return a

print("a antes de la mutacion", a)
a = addonethreetimesfunc(a)
print("a despues de la mutacion",a )
````

Ejemplo:

Como hacer una suma:
* con ciclos (bucles)
* recursivamente

````python
# con ciclos
# l es una lista
def sum(l):
    s = 0
    for  i in range(len(l)):
        s += l[i]
    return s

print(sum([1,3,5,4]))
````

````python
# recursivamente
def sumarecursiva(l):
    if len(l) ==  0:
        return 0
    else:
        if len(l) == 1:
            return l[0]
        else:
            return l[len(l)-1] + sumarecursiva(l[:-1])

print(sumarecursiva([1,3,5,4]))
````

````python
print(sumarecursiva([]))
````

````python
print(sumarecursiva([5]))
````
