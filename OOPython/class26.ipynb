### bloque `try--except``.

```python
def divideMe(a,b):
    try: # tratamos de dividir
        c = a/b # critical statement
        print(c)
        return
    except Exception:
        print("usted no puede dividir por 0")
        return

a=5
b=0
divideMe(a,b)
print("bye")
```

Capturemos el mensaje de error que arroja python en las exceptions

```python
def divideMe(a,b):
    try:
        c = a/b  
        print(c)
    except Exception as er:
        print(er)
        return
    return

a=5
b=0
divideMe(a,b)
print("bye")
```

En general, el print("bye") es solo por experimentar, pero en la practica se quieren hacer cosas donde se abren y cierran archivos, se crea y borra memoria.

```python
def divideMe(a,b):
    try:
        print("abrimos un archivo")
        c=a/b
        print("cerramos el archivo")
    except Exception  as er:
        print(er)
    return

a=5
b=3
divideMe(a,b)
print("bye")
```

directivas:

`try -- except -- else`

```python
def divideMe(a,b):
    try:
        print("abrimos un archivo")
        c = a/b # critical statement
    except Exception as er:
        print(er)
        print("cerramos el archivo")
    else:
        # blah-blah-bla
        print("cerramos el archivo luego de algun trabajo")
    return

a=5
b=0
divideMe(a,b)
print("bye")
```

Lo mas completo es:
`try--except--else--finally`

```python
def divideMe(a,b):
    try:
        print("abrimos un archivo")
        c = a/b
    except Exception as er:
        print(er)
    else:
        print("cerramos el archivo")
    finally:
        # algo que queremos hacer despues de todo esto, pase lo que pase
        print("Quiero hacer otras tareas no importa lo que pase arriba")
        print("por ejemplo cerrar el archivo")
    return

a=5
b=0
divideMe(a,b)
print("Bye")
```

Problemas que se pueden presentar y necesitan atencion.
* No se puede cerrar un archivo que no se ha abierto
* No se puede abrir un archivo dos veces
* No se puede cerrar un archivo dos veces

Estos problemas se pueden arreglar con "banderas" (flag)
Veamos un ejemplo

```python
def divideMe(a,b):
    flagOpen = 0


    
    try:
        print("abrimos el archivo")
        flagOpen = 1

        c = a/b
    except Exception as er:
        print(er)
    else:
        if flagOpen:
            print("cerramos el archivo")
            flagOpen = 0
    finally:
        print(" hacemos algun trabajo que halla que hacer")
        if flagOpen:
            print("cerramos el archivo")

    return
a=5
b=1
divideMe(a,b)
```

```python
b=0
divideMe(a,b)
```

## Assertions:
La syntax:
`assert Expression[, Arguments]`

Es para detectar errores, bugs en los codigos. 
Chequear tipos, valores, etc.

Ejemplo, sin mensaje de error

```python
def avg(speeds):
    n = len(speeds)
    assert n != 0
    return sum(speeds)/n

vels = []
print("Promedio de velocidades:", avg(vels))
print("Bye")
```
