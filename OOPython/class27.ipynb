Script to enable/disable touchpad.
```
#!/bin/bash
ID=$(echo `xinput list | grep Touchpad` | cut -d'=' -f 2 | cut -d' ' -f 1)
ENABLED=$(xinput --list-props $ID | grep Enabled | cut -d':' -f 2 | xargs)
if [[ $ENABLED == 1 ]]; then
  xinput -disable $ID
  echo Touchpad is disabled
else
  xinput -enable $ID
  echo Touchpad is enabled
fi
```

### Analisis de precedencia



````python
class Car:

    def __init__(self, model, year):
        self.model = model
        self.year = year
        return

    def __str__(self):
        return f'from str: My car is a {self.model} of the year {self.year}'

    def __repr__(self):
        return f'from repr: My car is a {self.model} of the year {self.year}'

c = Car("Toyota", 2012)
print(c)

        
````
