# Ejemplos de overlfow/underflow
Casos:
- entero (short half) de 16 bits sin signo, formato "h".
- entero (short half) de 16 bits con signo, formato "h".
- entero (int) de 32 bits sin signo, formato "i".
- entero (int) de 32 ts sin signo, formato "H".

[formatos de enteros con array](https://docs.python.org/dev/library/array.html)

```python
from array import array
a = array("h")
i=-2**(15)
a.append(i)

s = 2**(15)-1
a.append(s)
print("rango de numeros medios o cortos (short) es ",a)
```

```python
# creamos una situacion de overflow
a.append(2**15)
```

```python
a.append(-2**(15)-1)
```
