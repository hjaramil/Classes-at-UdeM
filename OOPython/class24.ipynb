## Scope: Alcance

```python
# variables de alcance local
def area_of_circle(r):
    pi=3.14159
    print("local variables\n", locals())
    return pi*r**2

print(area_of_circle(1))
```

```python
pi
```

```python
locals()
```

### Global Scope

```python
def area_of_circle(r):
    global pi # la hace global
    pi = 3.41415
    return pi*r**2

print(area_of_circle(1))
```

```python
pi
```

```python
globals()
```

```python
locals() == globals()
```

### Class Versus Object Scope

```python
class Cylinder:
    pi=3.14159

    print('\n locals() value inside class \n', locals())


    def __init__(self, radius, base, height):
        self.radius = radius
        self.base = base
        self.height = height
        return

    def volume(self):
        return pi*self.radius**2*self.height
```

```python
myCyl = Cylinder(5,2,3)
myCyl.__dict__
```

```python
myCyl.radius=7  # variable de instancia, objeto
vars(myCyl)
```

```python
myCyl.pi  # esta es variable de clase (scope class)
```

```python
# en una clase
class Circle:
    pi = 3.14159

    def __init__(self, r):
        self.r = r
        return

    def areaCircle(self):
        print("area of cirlce is", self.pi*self.r**2)
        return

```

```python
C= Circle(4)
C.areaCircle()
```

```python
vars(C) == C.__dict__
```

## Exceptions

Que tipos de errores se cometen cuando programamos.

1. Errores de complilacion (compile time errors). Se producen generalmente cuando la sintaxis es incorrecta. Por ejemplo olvidar los dos puntos luego de un `if`  o de un `for`  , problemas de indentacion. Estos son los mas faciles de corregir. 

2. Errores logicos (forman parte del run time errors pero los estuamos aparte). Estos son, tal vez, los mas dificiles. Un ejemplo



```python
a=b=c=1

c = a/(b+c)
d = a/b + c
c==d
```

```python
print(c,d)
```

3. Runtime erros. Errores del sistema por varias razones:\

    * Nan: Not a number. Division por cero, raiz de numero negativo, etc.
    * Segmentation Violation, Bus Error, Memory error, Segmentation Fault, etc. En C, errores de memoria. Pueden ser muy dificiles de depurar ("debugging"), muchas veces la memoria que esta por fuera de frontera se va "comiendo" el heap memory hasta inclusive invadir la parte donde se aloja el codigo y las variables . El problema cuando el error se arroja, ya es muy tarde. El problema ocurrio mucho mas a tras. A esto yo (Herman Jaramillo) lo llamo code metastasis. 

**Exceptions**: La excepcion es un evento que arroja un error o una advertencia acerca del programa que esta corriend. La exceptiones se pueden clasificar de varias formas de acuerdo al contexto. 

Por ejemplo, de acuerdo al estado del programa cuando corre.

1. **fatal error** : el programa para. 
2. **no fatal error** : el programa arroja un "warning" y siguie corriendo.

Otra clasificacion de la exepciones, por ejemplo division por cero, out-of-memory, invalud argument son.

1. Hardware failure (falla del hardware). Cuando falla un chip de memoria.
2. System error: Tratar de leer un archivo que no existe. Accesar memoria prohibida.
3. Software exception. Por ejemplo piense en unidades fisicas de volumen, masa y distancia. Deben ser numeros positivos. Si es negativo arroje (throw) un error.

### RecursionError

```python
def fac(n):
    if n < 0 :
        return "no defined for negative numbers"
    elif n <=1: 
        return 1
    else:
        return n*fac(n-1)

print("factorial of 3 is", fac(3))
```
