# Programacion orientada a objectos independiente del lenguaje

En la clase anterior hablamos de paradigmas de programacion.
Entre estos resaltamos la programacion estructurada, la programacion en procedimientos (procedural programming) y la programacion funcional.

El paradigma que se escoja para programar debe ser acorde con el problema que se quiera resolver. Por ejemplo piense en evaluar el siguiente polinomio

$$p(x)=1 + 3 x^2 + 2 x^3 - 4 x^4 $$

Calcular $p(x)$ para 
*  x=0.001
* x=0.5
* x= 3.2



```python
# para x=0.001
1 + 3* (0.001)**2 + 2*(0.001)**3 - 4*(0.001)**4

# hard coding (codigo duro)
```

```python
a=0.001
1 + 3* (a)**2 + 2*(a)**3 - 4*(a)**4
```

```python
# programacion estructurada
L=[0.001, 0.5, 3.2]
for i in L:
    print(1 + 3* (i)**2 + 2*(i)**3 - 4*(i)**4)
```

```python
# procedural programming 
def f(x):
    return 1 + 3* x**2 + 2*x**3 - 4*x**4

for i in L:
    print(f(i))
```

# Teoria de programacion orientada a objetos.
## Introduccion:

Que ventajas tiene el OOP.

* Modular: Divide y conquistaras. Codigos enormes en paradigma de procedimientos se pueden dividir en modulos indepndientes y faciles de manejar. Se necesitarian interefaces entre los modulos.  
* Simplificacion de codigo.  [Bjarne Stroutstrup](https://www.youtube.com/watch?v=uTxRF5ag27A) dice que los programas en programacion de procedimientos incrementan la complejidad con el cuadrado del numero de instrucciones, mientras que en OOP aumentan de forma lineal. Es decir para cosas pequenas es mejor "procedura" y para grandes OOP.  

* Recursos reusables (reciclable). Muchos objetos se pueden instanciar de una sola clase sin re-escribir la clase. Existen relaciones entre objetos como herencia y polimorfismo que tambien ayudan a que no halla que re-escribir codigo.

* Seguridad: En el sentido en que el codigo este protegido contra danhos causados por errores de programacion (bug). Y en el sentido en que clientes o competidores pueden mirar dentro de una clase y obtener informacion importante..

### Nota en el desarrollo de software en general.

Pensemos en un ejemplo practico de la vida real. La construccion de un puente. 

1. Analisis. Donde, por que, y  hacer planos. Arquitecto. Esta etapa es exploratoria.
2. Diseno. Aca entra el ingeniero. Aca se hacen los calculos.
3. Implementacion: Aca ya  entran los albaniles y los ingenieros dirigiendo y fisicamente se hace el  puente. En la parte de software,  aca es donde se escribe codigo. Es la programacion. 
4. Testing: pruebas. Las pruebas de puentes se pueden hacer cargandolos con un grupo de volquetas pesadas. Mejor que  sean autonomas. En el caso de desarrollo de software, un grupo de usuarios buscan como "quebrar" el codigo. 
Es mejor que los "probadores" encuentren los errores que luego un cliente se queje de que el software no trabaja (tiene "bugs"). "Los trapos sucios se lavan en casa".

5. Entrega (release). Cuando se corta la cinta.

Todos estos pasos pueden estar en ciclo..

* pre-alpha

### Razones para usar el paradigma OOP.

* Es **natural**: Se muestra el mundo como una coleccion de objetos.
* Es **confiable** : En varios sentidos. Recuerden la "seguridad". El mismo disenho lo hace confiable. 
* Es **reciclable** : Una clase puede "instanciar" muchos  objetos.  Existen relaciones como herencia y polimorfismo (explicamos mas adelante) que permiten que  el codigo se recicle facilmente.
* Es mas facil de "**debug**" (depurar). La separacion entre interface e implentacion y la division en clases hace que sea mas facil de depurar.

* Extensible: El OOP ayuda a la extensibilidad por su propio disenho. Una clase se puede extender (generalizar). Veremos como la herancia permite eso.

* Es flexible. Dos personas pueden trabajar en dos clases distintas del mismo paquete.
