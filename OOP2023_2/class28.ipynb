## Chain: continuacion
Este ejemplo lo encontre en StackOverflow.
El ejemplo consiste en hallar el "powerset" (el conjunto de subconjuntos de un conjunto), estos son $2^n$ (lo vimos en la clase de combinatoria)


````python
# uso de list comprehension para crear listas
import itertools
digits = [i for i in range(10)]
letters = [chr(value) for value in range(97, 123)]
capitals = [ch.upper() for ch in letters]


# contro de calidad
print("digits=",digits)
print("letters=",letters)
print("capitals=",capitals)
````

````python
from itertools import chain, combinations

def powerset(iterable):
    " powerset([1,2,3]) --> (), (1,), (2,), (3,), (1,2), (1,3), (2,3), (1,2,3)"
    s=list(iterable)
    return chain.from_iterable( combinations(s, r) for r in range(len(s)+1))

# ejemplo
myIter = ['a', 'b', 'c' , 'd', 'e']
A = powerset(myIter)
A = list(A)
print("los subconjuntos del conjunto {a,b,c,d,e} son", A)
print("el numero de subconjuntos de A es",len(A))
````

## ```compress```:
Se comporta como un filtro que selecciona datos de acuerdo a un selector. El comportamiento es como el de

Destras del telon pasa algo asi como esto.
```
def compress(data, selectors):
    # compress('ABCDEF', [1,0,1,0,11]) --> A C E F
    return (d for d, s in zip(data, selectors) if s)
```



````python
selectors = [1, 0, 1, 0, 1, 1]
letters = 'ABCDEF'
result = itertools.compress(letters, selectors)

for item in result:
    print(item, end='')
````

````python
selectors = [True, False, True, False, True, True]
letters = 'ABCDEF'
result = itertools.compress(letters, selectors)

for item in result:
    print(item, end='')
````

## ```dropwhile```:

Elimina elementos **mientras" el predicado sea verdad

## ```takewhile```:
sintaxis
```
itertools.takewhile(predicate, iterable)
```
Retorna elementos que satisfagan el predicado hasta que este se haga falso.

````python
L=[1,4,6,4,1]
result = itertools.takewhile(lambda x: x<5 , L)
print(list(result))
````

## ```acumulate```:
Sintaxis
```
itertools.accumulate(iterable[,, func, *, initial=None)
```

````python
# integrador (acumulador) clasico
# accumulate([1,2,3,4,5])-->1,3, 6, 10, 15
L=[1,2,3,4,5]
result=itertools.accumulate(L)
print(list(result))
````

````python
# otro ejemplos
result= itertools.accumulate(L, initial=100)
print(list(result))
````

````python
# cambiemos el operador que es "+" por defecto
import operator
# vamos a hacer multiplicacion en vez de suma, Productoria

L=[1,2,3,4,5]
result = itertools.accumulate(L, operator.mul)
print(list(result))

````

````python
# concatenemos
L=['1','2','3','4','5']
result = itertools.accumulate(L, operator.__concat__)
print(list(result))
````

# Generators in Python
Son como funciones, pero en vez de "return" usan "yield".
Lo que crean no son listas o tuplas o conjuntos sino iteradores (que son mas veloces y ocupan menos memoria).

````python
# por ejemplo los cubos
def cubesL(n):
    return [i**3 for i in range(n+1)]

cubesL(4)
````

````python
# usamos generadores
def cubes(n):
    yield [i**3 for i in range(n+1)]

gen = cubes(4)
print(gen)
print(type(gen))
print(list(gen))
````

````python
# otra forma de hacer lo mismo
def newCube(n):
    for i in range(n+1):
        yield i**3

m=4
gen = newCube(m)
print(gen)
````

````python
m=4
gen = newCube(m)
for i in range(m+2):
    print(next(gen))
````
