# Iterator Tools (Herramientas para iteradores)



````python
def areaHipo(a,b):
    area = a*b/2
    H = (a**2 + b**2)**(0.5)
    return area, H

a=3
b=4
(a2, H) = areaHipo(a,b) # 3,4,5 terna pitagorica.
print(f"area es {a2}, hipotenusa es {H}")
print(f"the area should be {a*b/2}, Hipotenusa should {(a**2 + b**2)**0.5}")
````

* ```count()```: Es contador infinito, comienza donde quiere, con itervalos fijos y llega hasta donde quiera.

* ```ziplongest()```: es una extension del ```zip```. Se basa en la lista mas larga para su evaluacion. Si tiene dos listas, se basa en la mas larga.

* ```cycle()```: Hace un ciclo, por ejemplo abcabcabcabc..... sobre un numero finito de objetos.

* ```repeat()```: crea objetos repetidos un numero de veces.

* ```starmap()```: es una extension de ```map()```, donde usa tuplas binarias en la forma $(x,y)$ para evaluar una funcion como $f(x,y)$.

* ```combinations()```: Crea las combinaciones (subconjuntos) de un conjunto. El conteo es  
dice que de $n$ elementos se pueden tomar
$$\binom{n}{k}$$
combinaciones.

* ```permutations()```:  retorna las permutaciones de $k$ elementos extraidos de un conjunto de $n$ elementos. La formula
es $n_k$ para el conteo. $n(n-1)(n-2) \cdots (n-k+1)$.

* ```products()``` : $k-$Productos cartesianos de $n$ elementos. El conteo $n^k$

* ```combinations_with_replacement()```: Aca puede haber repeticion.

* ```chain()```: Concatena dos o mas iteradores.

* ```islice()```: juega el papel de seccionar rebanadas en las listas

* ```compress()```: es un filtro que rechaza elementos de una lista de acuerdo a un selector. "Los son, se quedan"

* ```filterfalse()``` : Este es complemento del anterior. "Los que no son, se quedan".

* ```dropwhile()```: Elimina mientras. Mientras un predicado sea cierto elimina elementos. En el momento que el predicado sea falso, para y no se eliminan mas.

* ```takewhile()```: Este es el complemento del anterior, selecciona mientras el predicado sea cierto.

* ```acumulate()```: Es un integrador, sirve tambien para productorias, cuando el operador es producto en vez de suma.
$$ \prod_{i=1}^n a_i,   $$

## Combinatoria.
La regla basica, se llama la regla del producto.

Asuma que se tienen $k$ cajas y en la caja $i$ se almacenan
$b_i$ objetos. Cuantas posibilidades hay de escojer $k$ tuplas
con una tupla representada por un objeto por caja.






image.png


* Escojemos $b_1$ posibilidades de la caja 1
* Escojemos $b_2$ posibilidades de la caja 2
* $\vdots$
* Escojemos $b_k$ posibilidades de la caja k
El total de posibilidades es "la regla del producto"

$$\text{count} = b_1 \times b_2 \times \cdots \times b_k . \tag{1} $$

Que pasa si todas las cajas tienen el mismo numero $k$ de elementos. De la formula (1) con $b_i=k$ tenemos

$$\text{count} = n^k .$$
Para que sirve este modelo.

* $k$-tuplas. Piense en puntos de coordenadas
$(x_1, x_2, \cdots, x_k) \in \mathbb{R}^k$, entonces
Si cada coordenada puede tomar, exactamente $n$ valores se tiene
un total de $n^k$ valores.

##  sets:
Se tiene un numero finito de elementos sin repeticion y donde no importa el orden

Tomemos $k$ cajas, y tenemos $n$ elementos para distribuirlos en
las $k$ cajas. Con repeticion es el caso de productos $n^k$.
Sin repetir y donde el orden importa se llama

### Sin repetir pero importando el orden (permutacion)
importa el orden es:$ab \ne ba$.

no importa el orden es:$ab = ba$.

* Para la primera caja podemos escoger cualquiera entre $n$ posibilidades.

* Para la segunda caja podemos escoger cualquiera entre $n-1$ posibilidades. Pues ya la primera tiene uno que no se puede repetir.

* Para la tercera caja podemos escoger cualquiera entre $n-2$ posibilidades. Pues ya las dos primeras tiene escogido de a uno  que no se puede repetir.

* Asi sucesivamente. Para la caja $k$ solo disponemos de
$n-(k-1) = n-k+1$.

El modelo de las cajas me da que el conteo es:


$$\text{count} = n \times n-1 \times n-2 \times \cdots \times n-k+1 = \frac{n \times n-1 \times n-2 \times n-k+1 \times (n-k)!}{(n-k)!} = \frac{n!}{(n-k)!} = n^{\underline{k}}$$
Si $n=k$, entonces el coneo es $k!$ o $n!$.
Al numero $n^{\underline{k}}$ se le conoce como el
[falling factorial](https://en.wikipedia.org/wiki/Falling_and_rising_factorials).

ejemplos.
* Cuenta las permutaciones de $\{a, b, c \}$.
$(abc), (acb), (bac), (bca), (cab), (cba)$. $3! = 6$.
Por defecto, permutaciones es usando el mismo numero de elementos
del conjunto. O sea, en este caso, 3.

* Dado el conjunto $\{a,b,c \}$. Halle las permutaciones de 2 elementos. Cuentelas.
$n=3$, $k=2$, el conteo es:
$$ \text{count} = \frac{3!}{1!} = 3!=6 .$$
Serian $ab, ac, ba, ca, bc, cb$

* Otro ejemplo de $\{a,b,c,d \}$ permutaciones de a dos elementos.
$
$n=4$, $k=2$

$$\text{count} = \frac{4!}{(4-2)!} = \frac{4!}{2!}= 4 \times 3=12 .$$

Estas son: $ab, ac, ad, ba, bc, bd, ca,  cb, cd, da, db, dc $









Se tiene un numero finito de elementos con repeticion y no importa el orden.











### Sin repetir pero el orden no importa. $ab=ba$.
Si se tiene $k$ cajas, se tiene $k!$ permutaciones de esas $k$  cajas. Todas estas permutaciones se deben contar como una sola.
Entonces a la formula anterior, la debemos dividir por $k!$
y tenemos

$$\text{count} = \frac{n!}{(n-k)! k!} = \binom{n}{k} .$$
A estas se le llaman **combinaciones**. Es ta formula binimial la vamos a ver en otros contextos. Las combinaciones son el numero de subconjuntos de $k$ elementos de  un  conjunto de $n$ elementos.
