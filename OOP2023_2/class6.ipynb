# Scope (alcance)
Que es el alcance de una variable?
Depende: Estamos hablando de tiempo o de espacio o ....

Hay alcance temporal. Este es tiempo en que la variable permanece en memoria. Unas variables pueden permanecer en memoria todo el tiempo (desde que el programa aranca, hasta que termina). Una pequena fraccion (por ejemplo dentro de una funcion que corre en 1 milisegundo).

Hay alcance espacial. Una variable esta dentro de una funcion y el alcance es solo dentro de esa funcion no por fuera. Hay variables que estan en un archivo, y la variable puede tener maximo el alcance de ese archivo.  

Hay alcances

**local**: Se define dentro de una funcion.
**global**: puede estar dentro de una funcion pero se declara como  **global** en el sistema (veremos ejemplo).

**built-in scope**: Estas variables que ya estan metidas en el sistema y duran mientras el programa corre. Por ejemplo ```SyntaxError```.  Se puden listar las variables "built-in" en Python con el comando ```dir```




````python
dir(__builtins__)
````

Para que necesitamos clases u objetos si con lo que se me defiendo?

Por ejemplo, si yo voy definir el area de un cuadrado

````python
def sqr(x):
    return x**2

print(sqr(3))
print(sqr(4))
````

````python
# como se escribe este programa en OO
class Square:
    side = 2  # variable de clase por defecto

    def sqr(self):
        return self.side**2

# instanciemos la clase
mySqr = Square()
print(mySqr.side)
````

Que es **self**?
Por que es importante.
```self``` es la direccion del objeto. Cuando se crea un objeto, este tiene una direccion de memoria. Todo lo que se refiera a ese objeto tiene que estar ajustado (pegado) a esta direccion. El ```self``` es obligatorio en cada funcion. (no se tiene ```self``` y veremos ejemplos)


Veamos esto en la practica.

````python
class Square:
    side=2  # por defecto, default

    def sqr(self):
        print("address of the obejct", id(self))
        print()
        return self.side**2

mySqr = Square()
mySqr.sqr()
print(id(mySqr))
````

````python
mySqr = Square()
mySqr.sqr()
print(id(mySqr))
````

````python
class MyPower:
    side = 3

    def mypow(self, n):
        return self.side**n

myPower = MyPower()
myPower.mypow(5)
````

## Metodo ```__init__()```
El metodo ```__init__()``` es una funcion que se "dispara" cada vez que se crea un objeto.  En C++ y Java, a esto se le llama el ```constructor```.  En C++ y Java existe el ```distructor``` que destruye el objeto. En Python no hay ```desctructur``` (hay ```delete``` (esto lo vemos luego).

Hacemos un ejemplo del ```__init__()```, en la proxima clase mostramos el ```__new__()```

````python
class Car:
    def __init__(self, marca, modelo, anho):
        print(f'Initializating the car object')
        self.brand = marca
        self.model = modelo
        self.year = anho
        return # cada metodo debe tener un return
myCar = Car('Toyota', 'Corola', 2010)
print(id(myCar))



````
