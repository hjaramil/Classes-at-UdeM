# Otros contenedores
Podemos usar la directiva ```__contains__``` para probar si un objeto es contenedor o no.

````python
# veamos por ejemplo que los arreglos (arrays) son contenedores
import numpy as np
a = np.array([1,2,3])
membersOfa = dir(a)
if membersOfa.count("__contains__") :
    print("a is a container")

````

````python
print(membersOfa)
````

````python
# otra forma es la siguiente
import collections
if(isinstance(a, collections.abc.Container)):
    print("a is a container")
````

# Iterators (iteradores)
Contienen un numero contable de objetos. Se puede recorrer los
objetos mediante el magic ```__iter__()```.
Vamos a ver ```iterables```, estos se concectan con el dunder ```__next__```.
Enseguida vemos la diferencia entre iterables e iteradores.


````python
# iterables: Estos son objetos sobre los cuales se puede iterar o hacer ciclos.Por ejemplo
L=['a','b','c']

for l in L:
    print(l)
````

Como sabemos si un objeto es iterable? Usamos ```dir()``` y si dentro de los miembros aparece ```__iter__()``` es un iterable.
Veamos una funcion para determinar si un objeto es iterable.

````python
# determina si un objeto es iterable
def isIterable(myObj):
    membersOfObj = dir(myObj)
    if membersOfObj.count("__iter__"):
        print(f"Object {myObj} is an iterable")
    else:
        print(f"Object {myObj} is not an iterable")
    return

isIterable(L)
````

````python
# creamos la clase Range
class Range:
    def __init__(self, start, end, increment):
        self.value = start
        self.end = end
        self.increment = increment
        return

    def __iter__(self):
        return self

    def __next__(self):
        if self.value >= self.end:
            raise StopIteration
        current = self.value
        self.value += self.increment
        return current

# probamos
nums = Range(1, 10, 2)
for num in nums:
    print(num)




````
