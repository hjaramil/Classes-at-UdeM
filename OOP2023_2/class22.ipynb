Cual tiene mayor precedencia ```str()``` o ```repr()```?

````python
# lo averiguamos implementando un ejemplo
class Car:

    def __init__(self, model, year):
        self.model = model
        self.year = year
        return

    def __str__(self):
        return  f'from str: My car is a {self.model} of the year {self.year}'

    def __repr__(self):
        return  f'from repr: My car is a {self.model} of the year {self.year}'

c = Car("Toyota", 2012)
print(c)
````

* ```__del__()``` borra un objeto. Funciona para hacerle un "overload" al destructor.

````python
# ejemplo del uso de __del__()
class Exm:

    # inicializamos
    def __init__(self):
        print("ejemplo para ilustrar __del__()")
        return

    # sobrecarga de borrar
    def __del__(self):
        print("destructor para el objeto de la clase Exm, fue llamado")
        return

obj = Exm()
print(obj)
````

````python
del(obj) # se supone que debe imprimir el mensaje de arriba
````

````python
print(obj)
````
