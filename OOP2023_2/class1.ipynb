Que es un paradigma? Es el conjunto de metodologias subyacentes en un contexto cientifico. Desde el punto de vista de programacion es una forma de escribir y organizar codigo.

Hay un articulo, de quien introdujo el concepto **paradigma** en los lenguajes de computacion, que es Robert Floyd.

### Clasificacion de los paradigmas.

Podemos pensar en dos bloques grandes.

* Paradigma Imperativo.
* Declarativo.

#### Paradigma Imperativo.
Es el mas comun, es directo, la mayoria de los lenguajes que manejamos: C, C++, Fortran, Cobol, Java, Python se acomodan a este paradigma. El paradigma imperativo se enfoca en el "como" no en el "que".

Dentro del paradigma imperativo hay una clase de lenguajes.

#### Paradigma del lenguaje estructurado.

Para poder entender el lenguaje estructurado debemos visitar la historia y revisar la instruction **goto**.

El goto era una instruccion que permitia enviar el proceso de una linea o otra arbitraria (especificada por un rotulo).
El goto fue muy controvertido en los 60s y 70s, 80s, por que generaba codigos complejos llamados popularmente como **spagetti code**

### Procedural programming paradigm. Paradigma de programacion en procedimientos.

Surge la necesidad de evitar repeticion de codigo. Se crean los procedimientos, funciones, subrutinas (Fortran) las cuales factorizan los codigos para simplicarlos y hacerlos mas faciles de leer y administrar.  La programacion con procedimientos es como la estructurada pero con funciones.
Ejemplos de lenguajes con esta calidad son C, Fortran.


### Object Oriented Paradigm. Paradigma en programacion orientada a objetos

Quiero citar a Bjarne Stroutstrup. Que decia Bjarne.
"I discovered that the problems increased with the size of the program instead of the square of the size of the program" refiriendose a la programacion en OO.

Unas pocas palabras, por que el detalle viene despues.
Mientras que la programacion estructurada usa estructuras, en la OO usa clases. La clase es una extension de la estructura. La estructura es un conjunto de atributos y pare de contar. La clase es un conjunto de atributos y metodos. Ejemplo clase "Carro"

* Atributos:
    * Sustantivos: Silla, el motor, la direccion, las llantas, etc.
    * Adjetivos: blanco, grande, es veloz, es fuerte

* Metodos:
    * Verbos: corre hacia adelante, hacia atras, acelera, frena, vibra, etc.

Unos cuntos "bullets" de OO

* En OO los datos (atributos) y los metodos (funciones) vienen juntos en el mismo objeto.
* OO facilita la seguridad en los datos (data hidding). Se construye software seguro
* La informacion se puede reciclar muy facilmente.  De muchas formas
    * Defines una clase (blue print, una huella, un plano, o una patente: invento un carro que es un vehiculo para transportar personas, que tiene 4 puertas, 4 llantas, que tiene.....) de cual puedes "instance" instanciar tantos objetos como quieras.
    * Herencia: Puedes heredar de otras clases y evitar definir de nuevo todo lo que esas clases tienen.
    * Polimorfismo: Lo veremos mas adelante. Un mismo objeto puede comportarse de distintas formas de acuerdo al contexto.
