## herencia simple
Ejemplo

````python
class Person:

    def __init__(self, firstName, lastName, birthDay):
        self.firstName = firstName
        self.lastName = lastName
        self.birthDay = birthDay
        return


myPerson = Person("Bill", "Star", "06/21/83")
myPerson.__dict__
````

````python
class Student0(Person):
    def pleaseSit(self):
        print("por favor tome asiento")
        return
````

````python
p1 = Person("George", "Bleistein", "13/08/60")
s1 = Student0("Mary", "Ann", "20/10/76")
print(vars(p1))
print(vars(s1))
````

````python
s1.pleaseSit()
````

````python
p1.pleaseSit()
````

Creamos una clase que hereda de Persona y que no comparte
el metodo ```__init__```

````python

class Student(Person):

    def __init__(self, firstName, lastName, birthDay, grade):
        super().__init__(firstName, lastName, birthDay) # heredan del padre
        self.grade = grade
````

````python
myStudent = Student("Jorge", "Agudelo", "05/21/1980", 11)
myStudent.__dict__
````
