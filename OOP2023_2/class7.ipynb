## Encapsulacion:

Unos bullets que dijimos anteriormente.
Encapsulacion es poner las cosas en capsulas (clases)

Ventajas:

* Informacion hidding: Se puede ocultar informacion en una capsula.  Se puede ocultar la complejidad.

** Usted puede proteger sus propios datos.
** Protegerse de otros. En los ejemplos anteriormente vistos hablamos de variables privadas, que no pueden ver otras clases. Por ejemplo un peaton no deberia ver la temperatura de su carro.
Aunque las variables privadas permiten esconder informacion, Python le da la vuelta a esto. "Para python no hay nada oculto"
De esto verermos ejemplos.

* Modularidad. La implementacion de software por modules es mas clara y facil de mantener.

* Reuso. Las clases pueden generar tantos objetos como se quieran con basicamente el mismo codigo.


En el OO la informacion se oculta pero se crean iterfaces para poder obtenerla ("getters") o para modificarla ("setters").

Veamos ejemplos de como se puede "ocultar" informacion en una clase.











````python
class Car:
    def __init__(self, brand, model, year, engineTemperature):
        self.brand = brand
        self.model = model
        self.year = year
        self.__engineTemperature = engineTemperature

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1


test = Car("Toyota", "corolla", 2015, 79)
test.temperaturePlusOne()


print(test.brand)
print(test.model)
print(test.year)

print(test.__engineTemperature) # esto genera error
````

````python
test.temperaturePlusOne()
````

Como podemos accesar elementos privados de la clase

````python
class Car:
    def __init__(self, brand, model, year, engineTemperature):
        self.brand = brand
        self.model = model
        self.year = year
        self.__engineTemperature = engineTemperature

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1

    # agregamos metodos para obtener parametros que son privados (ocultos) para la case
    def getTem(self):
        return self.__engineTemperature

    # a esto se le llama un "setter". Es un parametro "mutable"
    def setTem(self, newTemp):
        self.__engineTemperature = newTemp
        return



test = Car("Toyota", "corolla", 2015, 79)
test.temperaturePlusOne()


print(test.brand)
print(test.model)
print(test.year)

#print(test.__engineTemperature) # esto genera error
t = test.getTem()
print(t)
````

La forma como se oculto la variable ```__engineTemperature``` se llama
[name manglinig](https://www.geeksforgeeks.org/name-mangling-in-python/)

Ver el link de arriba. Simplemente oculta el valor de la variable pero Python permite verlo para los usuarios que saben como.



````python
class Car:
    def __init__(self, brand, model, year, engineTemperature):
        self._brand = brand # esta variable es privada.
        self.model = model # publica
        self.year = year # publica
        self.__engineTemperature = engineTemperature #mangled variable

    def temperaturePlusOne(self):
        return self.__engineTemperature + 1

    # agregamos metodos para obtener parametros que son privados (ocultos) para la case
    def getTem(self):
        return self.__engineTemperature

    # a esto se le llama un "setter". Es un parametro "mutable"
    def setTem(self, newTemp):
        self.__engineTemperature = newTemp
        return



test = Car("Toyota", "corolla", 2015, 79)
test.temperaturePlusOne()


print(test._brand)
print(test.model)
print(test.year)

#print(test.__engineTemperature) # esto genera error
t = test.getTem()
print(t)
````

En Python se pueden definir variables privadas. Estas comienzaon con un "underscore" "_". Las variables no se ocultan sino que es un pacto entre programadores, no cambiar estas variables, ni verlas.



## Default parameters: Parametros por defecto

Veamos un ejemplo

````python
class Car:

    def __init__(self, brand='Nissan', model='Sentra', year=2015, engineTemperature=100):
        self.brand = brand
        self.model = model
        self.year = year
        self.__engineTemperature = engineTemperature

        return


    def temperaturePlusOne(self):
        return self.__engineTemperature+1

    def getTemp(self):
        return self.__engineTemperature

    def setTemp(self, newTemp):
        self.__engineTemperature = newTemp
        return


test=Car()

print("temperature + 1=", test.temperaturePlusOne())
print("brand", test.brand)
print("model", test.model)
print("year", test.year)

# imprima temperatura antes del cambio
print("temperature antes del cambio", test.getTemp())

# cambiar temperature a 110
test.setTemp(110)

# imprima temperatura despues del cambio
print("temperature despues del cambio", test.getTemp())




````

````python
class Car:

    def __init__(self, brand, model, year=2015, engineTemperature=100):
        self.brand = brand
        self.model = model
        self.year = year
        self.__engineTemperature = engineTemperature

        return


    def temperaturePlusOne(self):
        return self.__engineTemperature+1

    def getTemp(self):
        return self.__engineTemperature

    def setTemp(self, newTemp):
        self.__engineTemperature = newTemp
        return


test=Car("Nissan", "Sentra")

print("temperature + 1=", test.temperaturePlusOne())
print("brand", test.brand)
print("model", test.model)
print("year", test.year)

# imprima temperatura antes del cambio
print("temperature antes del cambio", test.getTemp())

# cambiar temperature a 110
test.setTemp(110)

# imprima temperatura despues del cambio
print("temperature despues del cambio", test.getTemp())




````
