En Unix el caracter "|" (pipe) concatena dos comandos. La salida de el primero es la entrada al segundo.
Ejemplo:
El comando
```cat california_housing_test.csv | wc```
funcion asi:

* El comando ```cat california_housing_test.csv``` lista el contenido del archivo a un buffer que usted nove.
* El buffer del primer comando para a ser la entrada al comando ```wc``` que cuenta las palabras (word count).

````python
cat california_housing_test.csv | wc
````
