# Contenedores: Containers
Son objetos que albergan un numero de objetos. Son grupos.

Hay varios de contenedores in ```Python```

## Built-in containers:
* ```list```:
    * Es mutable
    * No importa el orden
    * Esta indexada (por enteros)

* ```tuples```:
    * no son mutables
    * importa el orden
    * esta indexada por enteros

* ```dict```
    * se puede cambiar
    * Se accesa por llaves (keys)
    * no importa el orden

* ```set```
    * es mutable
    * importa el orden
    * no son indexados.


Hay otros tipos de contenedores en el modulo ```collections```.

Para verificar que un objeto es un ```container``` puede usar el magic ```__contains__```. Otras formas de preguntar si un objeto es contenedor es medaiante la pregunta

```
isinstance(a, collections.abc.Container)
```
Si es ```True``` es un contenedor. Ahora veremos ejemplos de esto.


Veamos un poco mas de detalle.
## Listas: (Lists)

* se pueden rebanar mediante el acceso de indices
* pueden mutar. Se puede insertar un elemento en cualquier parte de la lista.
* se puede iterar a lo largo de ella.


````python
myList = [1, 6, "hello", "abc", 2.3] # no es homogenea
````

Veamos como podemos con esta lista entender el "slicing"

Directivas generales:
La syntax is (sintaxis):

slicing has the syntax ```[start:end:increment ```

* ```increment```: si no se da, por defecto es 1. Puede ser negativo que nos vemos en "reversa"
* ```start```: si no se da es 0 (el primero)
* ```end```: Si no se da es el ultimo
* Para accesar el indice ```i```, se hace con la instruccion ```a[i]``` (donde ```a``` es el nombre de la lista


````python
myList[1]
````

````python
myList[2:4:2]  # slicing<
````

````python
myList[4:1:-1]
````

````python
# la lista vacia
myList = []
myList == list()
````

````python
# usar range
myList = list(range(10))
print(myList)
myList[-3] # desde atras hacia adelante, cuenta 3
````

````python
# ver todos los miembros de la clase lista
print(dir(myList))
````

Algunos metodos de los que se listan arriba son:

* ```append```: agrega  un elemento al final de la lista
* ```pop```: Elimina el ultimo elemento de la lista
* ```insert``` : inserta un elemento en la lista, usando un indice
* ```remove``` : remueve un elemento de la lista usando un indice
* ```sort``` : Ordena la lista, ascendente o descendente
* ```reverse``` : reversa el orden de la lista



````python
# ejemplo
myList = list([3,2])  # simplmente [3,2]
print("myList is", myList)
myList.append("hi")
print("appending 'hi' to my list results in ", myList)
myList.insert(10, "hello") # 10 esta muy lejos, lo pone al final
print("inserting 'hello' after the element 10", myList)
myList.remove("hi")
print("after removing 'hi' ", myList)
myList = [3,5,-1,7, 0]
print("cambio la lista a ",myList)
myList.sort() # ordena "in place"
print("luego de ordenarla es", myList) # de menor a mayor por defecto
myList = ["coco", "abc", "hello"]
print("la lista se sobre-escribe", myList)
myList.sort()
print("ordenada de abajo hacia arriba", myList)
myList. reverse()
print("en reversa es", myList)
````

````python
# ordenar de arriba a abajo
myList=[3,1,5,7]
myList.sort(reverse=True)
print("ordenada de arriba a abajo", myList)
````

## List Comprehension.
Sintaxis
```
newlist = [ expression for item in iterable if contion == True]
```

Aca
* ```expression``` : Es una funcion de ```item```.
* ```item```: Es la variable dependiente
* ```iterable```: es una lista, un ```range```, o otro iterable. Vemos iterables mas tarde.
* ```condition``` El resultado se escribe en ```newlist``` si la condicion es ```True``` solamente. Por ejemplo, si el iterable es ```i```, y quiere averguar si ```i``` es impar puede hacerlo de dos formas:
    * i%2 == 1 (forma larga)
    * i%2 (forma corta)

````python
# escriba los cuadrados los numeros impares entre 0 y 10
myList = [i**2 for i in range(10) if i%2==1]
print(myList)
````

En el ejemplo
* ```i**2``` es ```expression```
* ```i``` es el ```item```
* ```range(10)``` es el ```iterable```
* ```i%2==1``` es la condicion

````python
# escriba los cuadrados los numeros impares entre 0 y 10
myList = [i**2 for i in range(10) if i%2]
print(myList)
````

## Tuples:

Caracteristicas:
* Son inmutables (no se pueden cambiar). Es bueno por seguridad
* son ordenadas.
* Puen ser rebandas, como las listas.
* son utiles cuando queremos retornar varios valores de una funcion.
* Pueden ser usadas como llaves para diccionarios
* Son mas rapidas que las listas.
* Se notan con parentesis. Por ejemplo ```a=(1,3,5)```.

````python
a=(1,3,5,-2,4,8,9)
print(a[2])
````

````python
a[1:7:2]
````

````python
# los miembros de una tupla
print(dir(a))
````

````python
print(len(a))
````

````python
a.__sizeof__()
````

````python
a.index(5) # retorna el indice correspondiente al #5 en la lista
````

````python
print(a)
````

````python
# ejemplo donde se usan tuplas para retornar varios objetos
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        return

myStudent = Student('Robert', 21)
name , age = myStudent.name, myStudent.age
print("name=%s, age=%d\n"%( name,age) )
````

````python
# ejemplo donde se usan tuplas para retornar varios objetos
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        return

myStudent = Student('Robert', 21)
(name , age) = myStudent.name, myStudent.age
print(f"name={name}, age={age}\n")
````

````python
a,b,c=1,2,3
print(a,b,c)
````

````python
(a,b,c)=1,2,3
print(a,b,c)
````
