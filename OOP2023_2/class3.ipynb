# Object Oriented Theory
Nosotros hemos sobrevivido por anhos usando programacion imperativa, y los algoritmos han funcionado. Por que aprender un paradigma nuevo?

* **Modularidad y Naturalidad**: El OOP (Object Oriented Programming)
La programacion orientada a objetos es mas modular. No es tan "granulada".  Es una forma mas "natural" de resolver los problemas.

* **Simplificacion de Codigo**. Bjarne Stroustrup (en entrista en YouTube con Lex Friedman) dice: OO reduce el tamano de los programas, con respecto a los otros paradigmas, de forma lineal, mientras que los otros incrementan el tiempo de forma cuadratica.  Ver grafica a mano alzado.  

* **Complejidad** : La complejidad se puede ocultar en cada objeto. Por ejemplo un carro. Es un objeto muy complicado, pero nuestros lo ven como una figura con 4 llantas, 4 puertas y con cierta forma y tamano y que presta ciertas functiones. Pero un mecanico puede ver lo que esta "under the hood" (bajo capo). Y ahi hay otra complejidad (superior). Y un corredor de formula 1, lo ve de otra forma. Acelerador, freno direccion, y sus instintos y su capacidad de correr. Cada uno en su **nivel de abstraccion**. Usted puede esconder la complejidad para evitar problemas.



image.png

* Reusabilidad. Los objetos son 100% reciclables pero mas que eso. Se pueden extender a otros mediante herencias y polimorfismos. Esto lo vemos luego.


* Seguridad: El acceso a clases y obejtos puede ser restringido (y esto es bueno desde el punto de vista de seguridad). Vamos a ver que existen clases privadas, publicas.




## Historia:
El primer lenguaje que tuvo este paradigma fue SIMULA. Se desarrollo en los 1960s. Se desarrollo en Noruega: Norwegian Computing Center in Oslo. Por Ole-Johan Dahl y Kristen Nygaard, basado en parte en ALGOL-60

Bjarne Stroutstrup se refiere al siguiente problema.
"You do not want to write a city simulation with vehicles where you design a signal for a bicycle to turn to the left, if it is a car turn to the rightr, if it is a firetruck...."  Este tipo de problemas dio origen a lo que se llama herencia y polimorfismo. Donde no se necesita un arbol de "if statements".


Alan Kay. Trabajo en XEROX. Fue el inventor del GUI: Graphical User Interface. De aca salieron las ventanas del primer McCintosh (Steve Jobs) y las ventanas de Microsfot Windows-95 (antes era DOS). A Alan Kay se le atribuye el termino "object oriented".

## Teoria de OO.

### Una nota con desarrollo de software en general.
Pensemos, por ejemplo, en la construccion de un puente.

Que pasos hay?

* Analisis: Esta fase es exploratoria. Piense en la arquitectura que debe tener el puente (el arquitecto). No conocemos mas detalles de la construccion. Esta es fase de papel.
En la comparacion con un lenguaje, se pueden usar diagramas donde se hagan bosquejos de como debe funcionar el sistema. Por ejemplo existe una tecnica conocida como Unified Modeling Language.


* Diseno: Esta no esta totalmente separada del analisis. El ingeniero debe **calcular** la cantidad de cada material (ciencia de materiales) y como se preparar para que la estructura soporte los pesos, etc.  

* Implementacion: Es la construccion del puente. El arquitecto, ingeniero, los albaniles, etc construyendo el puente. En software development, esta es la implementaicion del algoritmo.


* Prueba: En el puente, se puede probar con un conjunto de volquetas cargadas pasando sobre el mismo. Si el riesgo humano se quiere evitar, se pueden usar volquetas autonomas.

En el desarrollo de software, las pruebas se hacen con programas que buscan "quebrar" es sistema.

* Entrega (Delievery): Se "corta la cinta" y queda listo para el uso del publico. El software se envia a los clientes.

Este ciclo, se puede repetir varias veces. De hecho existen varias etapas en la entrega de softare:


* Pre-alpha (la primera etapa con riesgos)
* Alpha (un poco mas robusta)
* Beta (todavia mas robusta)
* Release candidate (candidata a entrega)
* Stable Release (entrega estable)

La ultima lista de esta clase. Razones por las cuales OO es importante.

* Es **natural**. En el sentido en que navegamos por medio de objetos y no funciones o ecuaciones  o lineas.

* Es **reliable** (confiable). Cuando los programas se construyen con clases (son los mapas de los objetos)) la division de los objetos lo hace confiable, por poder identificar problemas mas facilmente

* Es **reusable**.  Los objetos se pueden usar una y otra vez en muchos ambientes, se pueden extener con herencia y polimorfismo etc.

*  Facil de mantener:  El soporte de un programa es el 80% y del desarrollo el 20%.  El disenho OO se presta para el mas facil mantenimiento de un programa.

* **Debuggin**. Cada objeto es como un mundo, y el objeto se debe probar y debe tener modo de ayudar en depuramiento del codigo. Al punto de que se puede detectar que el problema esta en ese objeto y centrar las fuerzas correctamente.

* **extensible**: Volmenos a lo de herencia y polimorfismo. Los objetos son facilmente extensibles.

* **flexible y modular**: Por modulos las cosas son mas faciles de administrar. Piense en la estacion es espacial. Tiene muchos modulos. Un modulo pertenece a un pais. Por ejemplo Brazil. Y ellos se encargan solo de ese modulo y las intefaces.

*  **hiding complexity**: revisar el ejemplo del telefono transparente. Poner las cosas en el nivel de abstraccion adecuado. Un carro: el chofer, el mecanico, el pasajero, el peaton.
