## Overloading (sobrecarga)
Vemos dos tipos de overloading.

### Function Overloading: sobrecarga
en una funcion. Este concepto trabaja bien en Java, C++ pero **no** en Python como lo vamos a ver.
"Hay que darle la vuelta", "walk around".


````python
class Sum:

    def __init__(self):
        return


    def sum(self, x, y):
        return x + y

    # En override el signature se conserva, en function overload el signature cambia
    def sum(self, x, y, z):
        return x+y+z

a = Sum()
print(a.sum(2,3,4))
````

````python
print(a.sum(2,3))
````

````python
# en python realmente pasa es "overwrite" (sobreescribe con el ultimo)
class Sum:

    def __init__(self):
        return


    # En override el signature se conserva, en function overload el signature cambia
    def sum(self, x, y, z):
        return x+y+z

    def sum(self, x, y):
        return x + y


a = Sum()
print(a.sum(2,3,4))
````

````python
print(a.sum(2,3))
````

Hay una forma de "darle la vuelta" al problema. Para esto necesitamos usar el "if" statement

````python
# veamos un ejemplo donde se sobrecarga el metodo __init__()
class Student():

    # first __init__()
    def __init__(self):  # no tiene parametros
        self.name = "Jack"
        self.score = 0
        return

    # second __init__
    def __init__(self, name, score):
        self.name = name
        if (score < 0 ):
            print("invalid score. It should be a positive number")
        else:
            self.score = score
        return

st = Student()


````

````python
# duck typing
class Sum:

    def __init__(self):
        return


    # En override el signature se conserva, en function overload el signature cambia
    def sum(self, x, y, z=0):
        return x+y+z

a = Sum()
print(a.sum(2,3))
````

````python
# como le damos la vuelta a la clase Student
# una forma puede user usando los parametros de default.
class Student2():

    # first __init__
    def __init__(self, *args): # variable arguments
        if len(args)==0:
            self.name = "Jack"
            self.score = 0
        if len(args)==2:
            self.name=args[0]
            if args[1] < 0:
                print("scores should be positive!")
            else:
                self.score = args[1]
        return

    def showData(self):
        print("name=%s, score=%d\n"%(self.name, self.score))

st = Student2()
st.showData()
````

## Operator overloading.
Los operadores,  por ejemplo, +, -, *, %, etc se pueden sobrecargar. El "duck typing" funciona bien y es una sobrecarga para muchos tipos. Pero que pasa si tengo libros y quiero sobre cargar el simbolo "+" para que sume las paginas. De forma que diga
LibroA + libroB= n
donde n es el numero de paginas de A + el numero de paginas de B.
Obvio que Python no sabe esto. Aca no hay "duck typing".


El overloading the "+" se hace con la funcion ```__add__()```
y el de multiplicacion es ```__mul__()```

````python
a=5
print(dir(a))
````

````python
# python tiene suma y multiplicacion de complejos, pero nosotros le vamos
# a hacer el "overloading", asumiendo que no sabemos
class Complex:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, otro): # add point to a point (x,y)
        x = self.x + otro.x
        y = self.y + otro.y
        return (x,y)

    def __mul__(self, otro): # multiplica el punto (x,y) por otro
        x = self.x*otro.x - self.y*otro.y
        y = self.y*otro.x + self.x*otro.y
        return (x,y)

````

````python
a = Complex(5,2)
b = Complex(-1,3)
print(a.__dict__)
print(b.__dict__)
````

````python
print("la suma a+b = ", a+b)
print("el producto a*b = ", a*b)
````
