# Naming Conventions
Ya habiamos hablado de esto, pero vamos a poner un poco mas de rigor.

La base de esto esta en
[PEP-8](https://peps.python.org/pep-0008/)

* Clases: Se sugiere que todas clases (nombres) comiencen con mayuscula y si hay palabras compuestas, cada palabra con mayuscula (camel-case, Pascal calse)
Por ejemplo ElectricCar, BabyGirl, VideoGame. Algunas veces usan "underscore" como Electric_car, Baby_girl, Video_game. El guion no se usa por que es caracter espcial para el signo menos "-". (la parte del guion bajo, no la veo en PEP-8)

* Atributos y metodos.  Comienzan con minuscula y cada palabra (compuesta) nueva con mayuscula (dromedary case).
Tambien se usan guiones bajos. Por ejemplo, colorType (color_type),
driveMode (drive_mode).

Un closure debe tener las siguientes caracteristicas.

1. Deben haber dos funciones anidadas (una dentro de otra)
2. La funcion anidada (la de adentro) defe referirse al argumento de la de afuera.
3. La funcion de afuera retorna la de adentro.

# Un tour por ```decorators```.

````python
def myMessage():
    print("hello")

def displayMessage(func):

    def inner():
        print("Excecuting", func.__name__, "function" )
        func()
        print("finished execution")

    return inner

testFunc = displayMessage(myMessage)
testFunc()
````

````python
# este mismo programa se puede escribir asi

@displayMessage
def myMessage():
    print("hello")

def displayMessage(func):

    def inner():
        print("Excecuting", func.__name__, "function" )
        func()
        print("finished execution")

    return inner

myMessage()
````

````python
# caso con multiples argumentos

def displayMessage(func):

    def inner(*args, **kargs):
        print("Executing", func.__name__, "function")
        return func(*args, **kargs)

    return inner


@displayMessage
def myMessage():
    print("Hello")
    print("finished executing")
    return

@displayMessage
def myMessageWithArgs(name, age):
    print("name", name, "age:", age)
    print("finished executing")
    return

myMessage()
myMessageWithArgs("Jorge",33)

````

# Decoradores con clases

````python
class DisplayMessage(object):

    def __init__(self, func):
        self.func = func
        return

    def __call__(self, *args, **kwargs):
        print("Executing", self.func.__name__, "function")
        return self.func(*args, **kwargs)

@DisplayMessage
def myMessage():
    print("Hello")
    print("finished execution\n")
    return

@DisplayMessage
def myMessageWithArgs(name, age):
    print("name", name, "age", age)
    print("finished execution")
    return

myMessage()
myMessageWithArgs("Jorge", 33)
````

````python
def divide(a,b):
    return a/b

a=5
b=0
print(divide(a,b))
````

````python
def smart_div(func):
    def inner(a,b):
        print("Dividing ", a, "by",b)
        if b==0:
            print("cannot divide by zero")
            return
        return func(a,b)
    return inner

@smart_div
def divide(a,b):
    return a/b

a=5
b=7
print(divide(a,b))
````
