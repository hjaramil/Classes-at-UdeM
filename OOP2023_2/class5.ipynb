# Inheritance (Herencia)
En la herencia tomamos como referencia dos clases. A una clase (de donde se hereda) la llamamos padre (parent) mientras que a la que hereda la llamamos hijo(a), **child**. A la clase padre tambien se le llama **superclass**, y la clase hija **subclass**.
Existen otros nombres: **base class** es la superclass y la **derived class** subclass.

* Parent, superclass, base class
* Child, subclass, derived class.

Donde se gana con esto. Herencia es una de las tecnicas mas eficientes para reciclar codigo (no re-escribir mucho).

Por ejemplo: Los **animales**, todos tienen una cantidad de propiedades en comun. Depronto las mas notorias es  que nacen, se reproducen y mueren. La idea es si tenemos un **mamifero**, basta con decir que es de la clase animal para evitar tener que repetir una cantidad de atributos y comportamientos. A otro nivel, un perro es un mamifero.  Cuando entremos a definir la clase **Perro** podemos asumir todas esas propiedades que ya los mamiferos tienen.

Un ejemplo elemental, usando algebra (elemental). Factorizacion

$$ a(x + y + z + w) = ax + ay + az + aw.$$
Desde el punto de vista algebraico, la expresion de la izquierda tiene 3 sumas y una multiplicacion. La derecha 3 sumas y 4 multiplicaciones. El ahorro cuando se desarrolla codigo es superior.

## Polimorfismo: (Polymorphism). Muchas formas.
Esta tambien es una forma eficiente de reclicar codigo.
Muchas veces objetos tienen comportamientos similares y una pequenha modificacion entre uno y otro hace la diferencia.

Hay varias formas de crear polimorfismo.

### Polimorfismo a traves de la herencia.
Pensemos en la clase animal. Podemos definir un metodo llamado **move** (moverse). Los animales se pueden desplazar. Parametros comunes a todos los animales en el desplazamiento (veremos ejemplos) son por ejemplo origen, destino, velocidad media.
Sin embargo una serpiente se mueve muy distinto a como se mueve un humano, un perro, un ave, un pez.  En ingles esto se llama **runtime polymorphism** o **dynamic polymorphism**.

Un tipo de polimorfismo es el **overloading** (sobrecarga).
Operator overloading es cuando la operacion (del operador) se hace de acuerdo al contexto. Si son enteros $a+b$ es la suma  de dos enteros, pero si son matrices $a+b$ es la suma de dos matrices.

Otro ejemplo (no lo tiene python). De acuerdo al numero de argumentos. Usted puede definir el metodo

```python
# overloading according to number of arguments
def sume(a,b):
    return a+b

def sume(a,b,c):
    return a+b+c

def sume(a,b,c,d):
    return a+b+c+d

a=5
b=6
sume(a,b)
```

```python
x=6
y=2
sume(a,b,x,y)
```

```python
# usted puede construir su propio overloading contando los argumentos.
def average(*args):
    return sum(args) / len(args)
print(average(999.0, 966.3, 988.2, 1344.5))

print(average(100,200))
```

Otra forma en que las clases se relacionan es mediante asociacion. Veamos un ejemplo tomado de pgf-nmled-manual.pdf

En los diagramas de UML para asociacion existen unas metricas que deben ser explicadas.

* 0..1: quiere decir que hay maximo hasta un objeto en ese lado.
* 0...n: Aca puede tener hasta n objetos o ninguno
* 1: Este quiere decir que tiene 1 y solo 1.
* 1...n: uno o mas.

Dentro de la asociacion hay dos tipos de asociaciones:

* **agregacion** (aggregation)
"has a": Por ejemplo en una canasta, en una finca, donde se guardan naranjas. La canasta tiene naranjas. A la agregacion se le llama tambien **containment**.
* **composicion** (composition):  "is part of" (es parte de).
Aca si la parte no puede existir sin el todo. Por ejemplo el corazon de un humano (latiendo) no puede existir sin el cuerpo.
