# Objectos y Clases (Objects and Classes)
## Objetos:
Los objetos como` en la vida real son cosas que peribimos, pero en la computacion toman una forma abstracta. Un carro en la calle es una y en la computacion. Solo que en la computacion es una representacion **abstracta** del objeto real. Esta representacion puede ser tan fiel como queramos. La descripcion del objeto es lo que llamamos **la clase**. El objeto es parte de un sistema, y de esa forma el sistema es una coleccion de objetos. El objeto tiene datos y comportamientos. Los datos (atributos) son lo que yo llamo adjetivos y sustantivos y los comportamientos (behaviors) son "verbos". El objeto es dinamico. Es decir hoy es una cosa y manana otra. Una foto de un objeto en el tiempo (snapshot) se llama el **estado** del objeto. Decimos entonces que el objeto tiene **atributos** y **metodos**. Estos pueden tener diferente visibilidad

* publicos (**public**): Cualquier objeto de afuera puede conocerlos y manipularlos (produce inseguridad).

* privados (**private**): Solo se pueden accesar por metodos dentro de la misma clase. Hay mas seguridad.

* protegidos: (**protected**): se pueden accesar por la misma clase o clases en la descendencia (el arbol genealogico)

Los objetos se comunican entre si mediante **mensajes**.
Los metodos invocan mensajes usando nombres y tipos.
Python es "typeless". Python aprende el tipo del contexto pero no hay especificarlos explicitamente.  Que es typeless? Por ejemplo
si $a,b$ son enteros, $a+b$ es una suma de enteros, pero si $a,b$ son matrices $a+b$ es una suma de matrices (eso es parte del OO). A esto se llama en argot popular **duck typing** (tipeado del pato). La idea es que en ingles "If it walks like a duck and it quacks like a duck, then it must be a duck". Es un polimorfismo por que el operador $+$ se adpta dinamiamente a los objetos sobre los que opera.

# Clases
Vamos a entender una clase por sus atributos.

* Una clase es un "blueprint" (plano)
* Es un tipo.
* La contiene un descripcion detallada de los atributos y metodos que los objetos deben tener. La clase es una abstraccion del objejo. El objeto es una "concrecion" de la clase.
* Cuando una clase se usa para crear un objeto, se dice que el objeto es un **instance** (intancia) de la clase. Inventamos, si no existe, el verbo "instanciar" cuando una clase genera un objeto. Veremos muchos ejemplo de esto.


## Acerca de la nominacion (convencion de nombres).
Las clases deben comenzar con mayuscula, y si la palabra es computuesta cada nueva palabra con mayuscula. A esto se le llama en ingles **CamelCase**. Por ejemplo ```AnimalKindom```.
Los metodos y objetos deben comenzar con minuscula, y cada palabra dentro del sistema compuesto a partir de esta con mayuscula. **dromedary case**. Tambien tambien se usan los terminos:

* Pascal Case <-> Upper Camel case
* Dromerary case <-> lower camel case

## Interaccion entre objetos.
La interaccion entre objetos por medio de mensajes que se pasan del uno al otro sirve para construir un sistema.
Para pasar un mensaje se requiere.

* El nombre del objeto
* Los parametros del objeto. (tipos)
* Typo de retorno (no para Python)





````python
def f(x):  # x parameter
    return x**2

f(5)  # 5 es argumento
f("hola")
````

La interface es una forma para informar al objeto de una comunicacion externa.

Existen muchas formas en que los objetos se pueden relacionar.
* relaciones de dependencia
* herencias (multiples, multinivel, etc.)

Hablemos de terminos importantes de las clases.

* constructor: El constructor es un **metodo**, este metodo se dispara automaticamente cuando se genera un objeto. Esto es de C++, Java, Python. Una vez se dispara el objeto el constructor **inicializa** todos los parametros que por defecto se definan.

* destructor: Cuando un objeto no se necesite mas, el destructor lo destruye (lo tira a la basura). Python no neceista destructores, por que usa "garbage collection".

* Metodos de comunicacion:
** setters: (escribir) Estos metodos son encargados de modificar o asignar nuevos valores (argumentos) a los parametros.
A los setters se les llama **mutators** (mutadores).
** getters: son metodos encargados de obtener (leer) informacio de los parametros en una clase.
