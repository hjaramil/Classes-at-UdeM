```python
# clase punto
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        return

    def draw():
        pass

    def __str__(self):
        return f"({self.x}, {self.y})"

    # sobrecarga de resta (overloading lo vemos despues)
    def __sub__(self, other):
        x = self.x - other.x
        y = self.y - other.y
        return(x, y)
```

```python
from numpy import sqrt
import matplotlib.pyplot as plt
import numpy as np

# clase segmento
class Segment():

    def __init__(self, p1, p2, color="blue"):
        self.p1 = p1
        self.p2 = p2
        self.color = color
        return

        # verificacion de que los argumentos son puntos
        if not isinstance(self.p1, Point):
            print("la firma (signature) es p1, p2, puntos de la clase , 'Point' ")
            return

        if not isinstance(self.p2, Point):
            print("la firma (signature) es p1, p2, puntos de la clase , 'Point' ")
            return

    def length(self):
        return ( (self.p2.x - self.p1.x)**2 + (self.p2.y - self.p1.y)**2)**0.5

p1 = Point(3, 4)
p2 = Point(2, 5)
s1 = Segment(p1, p2, "red")
print(s1.p2)
s1.length()

```

```python
# triangulo
class Triangle():

    def __init__(self, p1, p2, p3, color="blue"):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.color = color


        # construimos los lados del triangulo
        self.side1 = Segment(p1, p2) # composicion
        self.side2 = Segment(p2, p3) # composicion
        self.side3 = Segment(p3, p1) # composicion

        return

    def lengthSides(self):
        # (for cycle)
        print(f"lado entre {self.p1} y {self.p2} tiene longitud {self.side1.length()}")
        print(f"lado entre {self.p2} y {self.p3} tiene longitud {self.side2.length()}")
        print(f"lado entre {self.p3} y {self.p1} tiene longitud {self.side3.length()}")

    def perimeter(self):
        # la siguiente suma debe ser en un ciclo (blucle)
        return self.side1.length()+ self.side2.length()+ self.side3.length()

    def findArea(self):
        vect1 = self.p3 - self.p1
        vect2 = self.p2 - self.p1

        areaMatrix = np.array( [vect1, vect2])
        return 0.5*( abs(np.linalg.det(areaMatrix)))
```

```python
p1 = Point(1,3)
p2 = Point(3,4)
p3 = Point(-1,5)

t = Triangle(p1, p2, p3)
t.lengthSides()
```

```python
# clase Polygon:
class Polygon:

    # recibe lista de puntos
    def __init__(self, points, color):
        self.points = points
        self.color = color
        return

    # imprimir puntos
    def printPoints(self):
        print("The points for the polygon are:")
        for p in self.points:
            print(f" {p}", end='')
        return "    end"

    # dibujar poligono
    def draw(self):
        # disenho del titulo
        myStr = ""
        for p in self.points:
            myStr += f"{p}"
        title = f"Polygon for points {myStr}"
        plt.title

        # dos listas para abscisas y ordenadas
        x = []
        y = []
        for p in self.points:
            x.append(p.x)
            y.append(p.y)

        # toca pegar el primero (pare que cirre)
        x.append(x[0])
        y.append(y[0])

        # plot
        plt.plot(x, y, color=self.color) # lineas que unen los puntos
        plt.scatter(x, y, s=80, facecolors='none', edgecolors='k')
        plt.show()
        return


    def defPolygon(self):
        # saber cual poligono
        if len(self.points) == 2:
            s = Segment(self.points[0], self.points[1])
            print(f" The polygon is a {self.color} segment")
            print(f" The length of the segment is {s.length()}")
        elif len(self.points) == 3:
            t = Triangle(self.points[0], self.points[1], self.points[2])
            print(f" The polygon is a {self.color} triangle")
            print(f" The length of the sides are:")
            t.lengthSides()
            print(f" The are of the triangle is")
            print("        ", t.findArea())
        else:
            pass # tarea implementar 4latero y 5gono


        print(self.printPoints())
        print("Here is the figure of the polygone")
        self.draw()
        return






```
