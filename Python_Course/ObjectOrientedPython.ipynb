#### Intance Methods

Instance methods todos tienen self

```python
class Students():
    
    school = "Universidad de Medellin"
    
    # grades g1, g2, g3
    def __init__(self, name, g1, g2, g3):
        self.name = name
        self.g1=g1
        self.g2=g2
        self.g3=g3
        
    # promedio
    def average(self):
        return (self.g1 + self.g2+ self.g3)/3.0
    
# crear objetos
st1 = Students("Gloria", 3, 3.5, 3.2)
st2 = Students("Jorge", 4, 2, 1.0)

# imprima estudiante y promedio
print(st1.name + "   tiene" , round(st1.average(),2), "promedio")

print(st2.name + "   tiene" , round(st2.average(),2), "promedio")
```

### Accesors

```python
# se usan mas en java y en C++
# getters and setters: conseguir valores o asignar valores
class Students():
    
    school = "Universidad de Medellin"
    
    # grades g1, g2, g3
    def __init__(self, name, g1, g2, g3):
        self.name = name
        self.g1=g1
        self.g2=g2
        self.g3=g3
        
    # promedio
    def average(self):
        return (self.g1 + self.g2+ self.g3)/3.0

    def get_g1(self):
      return self.g1

    def get_g2(self):
      return self.g2

    # setters
    def set_g1(self, value):
      self.g1 = value
    
# crear objetos
s1 = Students("Gloria", 3, 3.5, 3.2)

# imprima estudiante y promedio
print(st1.name + "   tiene" , round(st1.average(),2), "promedio")

s1.get_g1()
s1.set_g1(4.0)
s1.get_g1()
```

```python
s1.g1=3
print(s1.g1)
```

## Metodos de clase

```python
# class methods
# A diferencia de los de intancia no necesitan el self 


class Students():
    
    school = "Universidad de Medellin"
    
    # grades g1, g2, g3
    def __init__(self, name, g1, g2, g3):
        self.name = name
        self.g1=g1
        self.g2=g2
        self.g3=g3
        
    # promedio
    def average(self):
        return (self.g1 + self.g2+ self.g3)/3.0

    @classmethod
    def info(cls):
      return cls.school
    
# crear objetos
s1 = Students("Gloria", 3, 3.5, 3.2)
```

```python
s1.info()
```
