## Teorema del Circulo de Gershgorin (repaso)
Cada $\lambda$ esta en algun circulo de Gershgorin

Primera consecuencia. Si la matriz $A$ es estrictamente dominante (por filas o culumnas), entonces es invertible. Asumamos que es estrictamente dominante por filas $| \lambda - a_{ii} | < R$ donde $R = \sum_{i \ne } | a_{ij} |$. Si $\lambda=0$


$$ |a_{ii} | < R$$
, lo que indique la matriz no es estrictamente dominante por filas.

O sea que $\lambda \ne 0$ y esto para cualquier $\lambda$ . O sea que todos los $\lambda$ son distintos de cero y la matriz es invertible.



El problema mas importante de estas primeras lecciones es

$$\bmx_i = T \bmx_{i-1} + \bmc .$$

Si el $\lim_{i \to \infty} \bmx_i = \bmx$ entonces podemos escribir

$$\bmx = T \bmx + \bmc .$$

De forma que $(I-T) \bmx = \bmc$.
Si el inverso  de $I-T$ existe


$$\bmx = (I-T)^{-1} \bmc .$$

Una identidad que debemos memorizar el algebra
)
$$\frac{1 - t^n}{1 - t} =1 + t + t^2 + \cdots + t^{n-1} .$$
Aca llegamos por otro lado


\begin{eqnarray}
I - T^2 &=& (I-T)(I+T) \\
I - T^3 &=& (I -T) (I + T + T^2) \\
I - T^4 &=& (I -T) (I + T + T^2 + T^3) \\
&\vdots& \\
I - T^n &=& (I-T) ( I + T + T^2 +\cdots + T^{n-1}) \tag{1}
\end{eqnarray}

Es decir
$$\frac{I - T^n}{I - T} =1 + T + T^2 + \cdots + T^{n-1} .$$

En Ecuacion (1) tomamos el limite cuando $n \to \infty$,
si $T$ converge $\lim_{i \to \infty} T^i = 0$
Entonces

$$ I = (I-T) ( I + T + T^2 + \cdots + T^n + \cdots ) .$$



Podemos decir que

$$ (I- T)^{-1} = \frac{I}{I-T} = I + T + T^2 + \cdots T^n + \cdots $$


Teorema: Asuma que la matriz $T$ de la recursion $\bmx_i = T \bmx_{i-1} + \bmc$, El radio espectral $\rho(T) < 1$ si y solo si $(I-T)^{-1}$ existe.


**Prueba**: $⇒$:

Asumimos que $\rho(T) < 1$, y sea $\lambda$ un autovalor de $T$. De forma que $T \bmx = \lambda \bmx$. Restando $\bmx$ a ambo lados
$(I-T) \bmx = (1 - \lambda) \bmx$. De forma que $1 -\lambda$ es autovalor de $I-T$. Ahora como $\rho(T) < 1$ entonces

$$1 - \lambda \ne 0  .$$
Probamos anteriormente que autovalores no cero corresponden a matrices invertibles. En este caso la matriz $I-T$ tiene inversa.

Vamos a recordar la recursion

\begin{eqnarray}
 \bmx_i &=& T \bmx_{i-1} + \bmc \\
 &=& T (T \bmx_{i-2} + \bmc) + \bmc \\
 &=& T^2 \bmx_{i-2} + (T + I) \bmc \\
 &=& T^3 \bmx_{i-3} + (T^2 + T + I) \bmc \\
 &\vdots& \\
 &=& T^i \bmx_0  + (T^{i-1} + T^{i-2} + \cdots + I) \bmc
\end{eqnarray}

Dado que $\rho(T) < 1$ entonces $\lim_{i \to \infty} T^i \bmx_0 =0 $.

De forma que

$$\bmx = \lim_{i \to \infty} \bmx_i = (I + T + T^2 + \cdots + T^n + \cdots) \bmc .$$

Es decir la recursion $\bmx_i = T \bmx_{i-1} + \bmc$ converge.


Prueba implicacion en sentido opuesto.
$⇐$: Se asume que $I-T$ tiene inverso y probamos que $\rho(T) < 1$.

Dato que $(I-T)^{-1}=I + T + \cdots + T^n + \cdots$ entonces

$\lim_{n \to \infty} T^n = 0$ por que si no entonces la serie diverge. Entonces $T$ converge y ya vimos en clase anterior que esto equivale a que $\rho(T) < 1$.

Tenemos que la convergencia de $I-T$ es equivalente a que el radio espectral $\rho(T) < 1$. Si $\rho(T) > 1$ hay divergencia, pero que pasa si $\rho(T) = 1$?

Asuma por ejemplo que $T=I$.


$\bmx_i = T \bmx_{i-1} + \bmc$

Si $T$ es la identidad $\bmc=0$ y
$\bmx_i =  I \bmx_{i-1} = \bmx_0$
como $\bmx_0$ es arbitrario el sistema tiene un numero infinito de soluciones. $\bmx = T \bmx$, $T=I$ cualquier $\bmx \in \mathbb{R}^n$ es solucion.

Tenemos condiciones para convergencia. Pero queremos estimar el error y la rapidez con que una recursion converja.

El error es $\epsilon = \| \bmx - \bmx_i \|$.

Vamos a calcular el error con las ecuaciones que obtuvimos hace unos minutos.

$$\bmx =  (I + T + T^2 + \cdots + T^n + \cdots) \bmc .$$
Entonces



\begin{eqnarray}
\bmx - \bmx_i &=&  (I + T + T^2 + \cdots + T^n + \cdots) \bmc
 - T^i \bmx_0 - (I + T + \cdots T^{i-1}) \bmc \\
 &=& (T^i+ T^{i+1} + \cdots + T^{i+k} + \cdots ) \bmc - T^i \bmx_0 \\
 &=& T^i(I + T + \cdots + T^k + \cdots ) \bmc - T^i \bmx_0 \\
 &=& T^i( \bmx - \bmx_0).
\end{eqnarray}

Para calcular el error tomamos la norma a ambos lados


$$\| \bmx - \bmx_i \| = \| T^i(\bmx - \bmx_0) \| \le \| T \|^i \| \bmx - \bmx_0 \| .$$


Esta ecuacion nos dice dos cosas:

* Si $\rho(T)$ es pequenho, el error es pequenho.
* Si $\bmx_0$ esta cerca a la solucion $\bmx$ la convergencia es rapida.

El problema con este sistema es que nosotro no conocemos $\bmx$.

Vamos a buscar otra forma de calcular el error sin necesidad de saber la solucion.


\begin{eqnarray}
    \| \bmx_{i+1} - \bmx_i \| &=& \| T \bmx_i - T \bmx_{i-1} \| \\
    &=& \| T( \bmx_i - \bmx_{i-1} ) \|  \\
    &\le& \| T \|  \| \bmx_i - \bmx_{i-1} \| \\
    &\le& \| T \|^2 \| \bmx_{i-1} - \bmx_{i-2} \| \\
    & \vdots & \\
    &\le &  \| T \|^i \| \bmx_1 - \bmx_0 \|
\end{eqnarray}

Vamos a usar dos cosas:

* Desigualdad triangular $\| \bmx + \bmy \| \le \| \bmx \| + \| \bmy \|$.

* La telescopica
$$a_n - a_0 = a_n - a_{n-1} + a_{n-1} - a_{n-2} + a_{n-2} - \cdots  -a_1 + a_ 1 + a_0 $$


Veamos


Asumamos $i > j$
\begin{eqnarray}
\| \bmx_i - \bmx_j \| &=& \| \bmx_i - \bmx_{i-1} + \bmx_{i-1} -
\bmx_{i-2} + \bmx_{i-2} + \cdots -\bmx_{j+1} + \bmx_{j+1} - \bmx_i \| \\
&\le& \| \bmx_i - \bmx_{i-1} \| + \| \bmx_{i-1} - \bmx_{i-2} + \cdots + \| \bmx_{j+1} - \bmx_j \| \\
&=& ( \| T \|^{i-1} + \| T \|^{i-2} + \cdots + \| T \|^j) \| \bmx_1 - \bmx_{0} \| \\
&=& \| T \|^j \left ( 1 + \| T \| + \cdots + \| T \|^{i-j-1}   \right ) \| \|  \bmx_1 - \bmx_0 \| \\
&=& \frac{ \| T \|^j (1 - \| T \|^{i-j})}{1 - \| T \|} \| \bmx_1 - \bmx_0 \|
\end{eqnarray}


# Aplicaciones espeficicas de la teoria de ecuaciones iterativas.

* Jacobi
* Gauss-Seide
* Metodos de relajacion (Over-relaxation methods)


La idea de estas tecnicas es escibir la matriz $A$
como la suma de tres matrices

$$A = L + D + U .$$
donde

* $L$ es triangular inferior. $L_{ij}=0$ $i \le j$.
* $D$ es diagonal. $d_{ij}=0$, $i \ne j$
* $U$ es triangular superior. $u_{ij}=0$, $i \ge j$

### Metodo de Jacobi.


\begin{eqnarray}
A \bmx &=& \bmb \\
(A + D + U)  \bmx &=& \bmb \\
D \bmx &=& -(L + U) \bmx + \bmb
\end{eqnarray}

De aca que


$$\bmx = -D^{-1}(L+U) \bmx + D^{-1} \bmb .$$


Pensamos en el problema de recursion donde
$$ T_J = -D^{-1}(L + U)  \quad , \quad  \bmc= D^{-1} b$$.

Y resolvemos el problema
$$ \bmx_k = T_J \bmx_k + \bmc .$$

Podemos escribir esta ecuacion en forma "granulada" (con indices).


$$ x_i^{(k)} = \frac{1}{a_{ii}} \left
[ \sum_{j=1, j \ne i}^n (-a_{ij}) x_j^{(k-1)} + b_i  \right ] $$

Para que el metodo de Jacob converja necesitamos que
$\rho(T) = \rho [ D^{-1}(L+U) ] < 1$.

**Teorema 2:** : Si $A$ es estrictamente diagonal dominante por filas (por columnas) las iteracion es de Jacobi relacionadas con $A$ convergen.

Prueba: Si es diagonalmente dominante

$$\sum_{j \ne i} | a_{ij} |_ < | a_{ii} | .$$
para cada $i=1,2, \cdots, n$ De forma que


$$ \frac{\sum_{j \ne i} | a_{ij} |}{| a_{ii}|} < 1 \quad , i=1,2_, \cdots, n .$$

Por lo tanto

$$\max_{i}   \frac{\sum_{j \ne i} | a_{ij} |}{| a_{ii}|} < 1 $$

El operador $T$ compule $T = -D^{-1}(L + U)$, o en forma granulada

$$T_{ij} = -a_{ij}/a_{ii} , i \ne j$$,
$$T_{ii} = 0 $$

$$ \| T \|_{\infty} = \max \frac{\sum_{j \ne i} | a_{ij}|}{| a_{ii}|}  < 1.$$

Habiamos visto que $\rho(A) \le \| A \|$, entonces

$\rho(T) < 1$.  O sea que el sitema converge.

Implementamos el ejemplo que hicimos a mano en `Python`

```python
import numpy as np
from numpy import linalg as LA

def Jacobi(A,b,x0,tol,maxiter):
    n = len(b)
    D = np.diag(A)

    # resto la diagonal de A
    LU = A - np.diag(D)
    x = np.zeros(n)

    # iteraciones
    k=0
    error = 10

    while ( k < maxiter and error > tol ):
        k += 1
        xnew = (b - np.dot(LU,x))/D
        error = LA.norm(xnew-x)
        x = xnew

        print(f"solucion x para la iteracion k={k} con error {error}")
        print("  ", xnew)
        x = xnew

    if k==maxiter:
        print("maximo numero de iteraciones alcanzado")
    else:
        print("solucion encontrada")

    return

## Prueba
A = np.array([[5, -2, 3],[-3,8,1],[-3,-1,-6]])
b = np.array([-1, 4, 0])
nmax=100
tol=0.001
x0=np.array([0,0,0])


# llamamos a Jacobi
Jacobi(A,b,x0,tol,nmax)
```

```python
# Verificamos la solucino con Python
np.linalg.solve(A,b)
```

```python
# implementamos el radio espectral
def radio_espectral(A):
    n = np.shape(A)[0]

    # extraemos la matriz diagonal
    D = np.diag(A)
    # restamos
    LU = A - np.diag(D)
    T = -np.matmul(LA.inv(np.diag(D)), LU)
    autovalores = LA.eigvals(T)
    print("T=")
    print(T)
    print("autovalores", autovalores)
    radio_espectral = np.max(np.abs(autovalores))
    print("radio espectral", radio_espectral)

    return


radio_espectral(A)
```
