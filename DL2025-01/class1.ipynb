## Transformaciones multilineales
Vamos a pensar en transformaciones lineales $T(x)=y$, donde
$x,y$ estan en algun espacio $T$ en otro.  Comenzamos de abajo hacia arriba.

* $x \in \mathbb{R}, y \in \mathbb{R}, a \in \mathbb{R}$.
La transformacion lineal mas sencilla

$$ Tx = y = ax$$
Esta transformacion corresponde a una linea recta que pasa por el origen. Eo objeto $y$ es un **tensor de rango 0** (escalar, no hay indices).

* $x \in \mathbb{R}^n, y \in \mathbb{R}^m$
$$ Tx = y = ax.$$
Donde cae $a$? Para que esto sea dimensionalmente correcto, necesitamos que

$$a \in \mathbb{R}^m \times \mathbb{R}^n = \mathbb{R}^{m \times n}$$
Analisis dimensional en Python

```python
import numpy as np
A = np.array( [  [2 ,3, 4], [1, 4, 5]])
B = np.array( [  [2 ,3 ], [1, 4]])

# multiply A times B matrix product
np.matmul(A,B)
```

Decirmos que $y$ es un **tensor de rango 1**.

Hagamos un analisis un poco mas profundo de esta operacion

La operacion $T$ es lineal
$$T( c_1 x_1 + c_2 x_2) = c_1 T(x_1) + c_2 T(x_2)  $$
Escribamos los vectores base canonicos

$$
e_1 = \begin{pmatrix} 1 \\ 0 \\ \vdots \\ \vdots \\ 0  
\end{pmatrix}
\quad, \quad e_2 = \begin{pmatrix} 0 \\ 1 \\ \vdots \\ \vdots \\ 0  \end{pmatrix}
\quad, \quad e_k = \begin{pmatrix} 0 \\ 0 \\ \vdots \\ 1 \\ \vdots \\ 0  
 \end{pmatrix}   \quad \cdots  \quad
 e_n =
 \begin{pmatrix}
 0 \\ \vdots \\ \vdots \\ 0 \\ 1
 \end{pmatrix}
 $$

 Podemos escribir $x = \sum_{i=1}^n x_i e_i$
 de forma que

 $$T x = T \left ( \sum_{i=1}^n x_i e_i \right ) = \sum_{i=1}^n x_i T(e_i).$$
 Encontramos que cada $T(e_i) es una columna de la matriz

 $$a = \begin{pmatrix}
 \vert & \vert & \vdots & \vdots & \vert \\
 T(e_1)  & T(e_2) & \vdots & \vdots & T(e_n) \\
 \vert & \vert & \vdots & \vdots & \vert \\
   \end{pmatrix}   $$


* $x \in \mathbb{R}^{n \times m}, y \in \mathbb{R}^{p \times q}$
y
$$Tx = y = ax  $$
Quien es $a$?

De la misma forma que hicimos en el ejercicio anterior creamos los vectores base para el espacio de salida $\mathbb{R}^{n \times m}$. Para entender esto veamos un ejemplo donde $n=m=2$.

Los "vectores" (son elementos de un espacio vectorial, pero realmente son matrices) canonicos son ahora

$$
e_{11} = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix} \quad , \quad
e_{12} = \begin{pmatrix} 0 & 1 \\ 0 & 0 \end{pmatrix} \quad , \quad
e_{21} = \begin{pmatrix} 0 & 0 \\ 1 & 0 \end{pmatrix} \quad , \quad
e_{22} = \begin{pmatrix} 0 & 0 \\ 0 & 1 \end{pmatrix} \quad , \quad
$$





Vamos a definir el **producto tensorial** (outer product).

Sean dos vectors $a$,$b$ con componentes $a_i$, $b_i$, entonces definimos el producto **tensorial** (outer product) como

$$ c_{ij} = (a \otimes b)_{ij}  = a_i b_j $$

Un ejemplo, $e_{11} = e_1 \otimes e_1$ pues

$$e_{11} = e_1 \otimes e_1 =
\begin{pmatrix} 1 \times 1 & 1 \times 0 \\ 0 \times 1 & 0 \times 0   
\end{pmatrix}
= \begin{pmatrix} 1 & 0 \\ 0 & 0  \end{pmatrix}  $$
Igual se puede probar que $e_{ij} = e_i \otimes e_j$, $i,j=1,2$.

Esto se puede generalizar a cualquier dimension. Sea $e_i$ un vector tal que tiene un $1$ la dimension $i$ y 0 en las demas dimensiones. Entonces el objeto $e_i \otimes e_j$  es la matriz que tiene un $1$ en la posicion $ij$ y 0 en las demas.





image.png

Podemos escribir

$$x = \sum_{i=1}^n \sum_{j=1}^m x_{ij} e_{ij} = \sum_{i=1}^n \sum_{j=1}^m x_{ij} e_i \otimes e_j $$.
Como $T$ es lineal

$$ T(x) = \sum_{i=1}^n \sum_{j=1}^m x_{ij} T(e_{ij}) \tag{1} $$

El objeto $T_{ij}$ para $i,j$ fijos es una matriz en $\mathbb{R}^{p \times q}$

En este caso el objeto $a$ tiene la forma

$$a_{st}^{ij} = T_{st}(e_{ij}) $$

En este caso $y$ es un **tensor de rango 2**

Escribimos $T$ (de (1)) usando sus componentes como

$$T_{st}(x) = \sum_{i=1}^n \sum_{j=1}^m x_{ij} T_{st}(e_{ij}) = \sum_{i=1}^n \sum_{j=1}^m a_{st}^{ij} x_{ij} $$


$$T_{st}(x) = x_{ij} T_{st}(e_{ij}) = a_{st}^{ij} x_{ij} $$
Esta expresion es la **notacion de Einstein**. Einstein dijo, para evitar tantas sumas (sigmas $\Sigma$) no las escribamos pero **ojo**, indices repetidos (dummies) se suman, los otros (libres) no.

Tenemos la forma

$$T(x) = a y $$

Generalizacion: Pensemos en

$$x \in \mathbb{R}^{n_1 \times n_2 \times \cdots \times n_j} $$
$$y \in \mathbb{R}^{m_1 \times m_2 \times \cdots \times m_k} $$

Queremos averiguar quien es $a$ en la transformacion

$$ T(x) = ay $$
que es lineal.

Sea $e_{p_1 p_2 \cdots p_j}=e_{p_1} \otimes e_{p_2} \otimes \cdots \otimes e_{p_j}$ donde $e_{p_j}$ tiene un 1 en la componente $p_j$ y 0 en todas las otras componentes.  La "super matrix" $e_{p_1 p_2 \cdots p_j}$ tiene 0 en todas las componentes **exepcto* en aqueyya con indices $p_1, p_2, \cdots, p_j$.

Usando la notacion de Einstein escribimos

$$ x = x_{p_1 p_2 \cdots p_j} e_{p_1 p_2 \cdots p_j} $$
De forma que el operador $T$ es

$$T(x) = x_{p_1 p_2 \cdots p_j} T(e_{p_1 p_2 \cdots p_j}) $$
$T$ no es un escalar. Faltan los indices de $T$
Escribamoslos

$$\left [ T(e_{p_1 p_2 \cdots p_j}) \right ]_{q_1 q_2 \cdots q_k} = a_{q_1 q_2 \cdots q_k}^{p_1 p_2 \cdots p_j} $$
No lo he dicho pero

$0 \le p_1 \le n_1$ (enteros
$0 \le p_2 \le n_2$ (enteros
$\vdots$
$0 \le p_j \le n_j$ (enteros)$0 \le p_1 \le n_1$ (enteros
$0 \le q_1 \le m_1$ (enteros
$\vdots$
$0 \le q_k \le m_k$ (enteros)

La transformacion la podemos escribir como

$$ T_{q_1 q_2 \cdots q_k(x) = a_{q_1 q_2 \cdots q_k}^{p_1 p_2 \cdots p_j}} x_{p_1 p_2 \cdots p_j}. $$




Decimos que $T_{q_1 q_2 \cdots q_k(x)}$ es un **tensor de rango** $k$
