## Redes Neuronales Artificiales. ANN: Artifical Neural Networks

Por que es importante. Existe un teorema matematico. [Universal Approximator](https://en.wikipedia.org/wiki/Universal_approximation_theorem .

Con numero suficiente de capas, podemos apromaximar (casi) cualquier funcion
de $f: \mathbb{R}^n \to \mathbb{R}^m$ siempre y cuando la profundidad de la red sea por lo menos $d= \max \{ n+1 , m \}$. Casi: INcluye todas las funciones continuas, integrales. En particular piensen en una imagen en el computador.
La imagen es una function

$$f : \mathbb{R}^n \times \mathbb{R}^m \to \mathbb{R} $$

### Historia:
1. 1943 : Warren McCulloh (neurofisiologo) y Walter Pitts (matematico).
Crearon la neurona logica (que la explicamos mas adelante). Fue el primer intento de comparar las neuronas biologicas con las neuronas artifiales (inteligencia artifical).

2. 1949: Donald Hebb escribio un libro "Organization of Behavior". 
Lo que el decia es que las conexiones entre neuronas tiene pesos y que estos
pesos se debilitan cuando la experiencia es negativa (por ejemplo fracasos). 
De otro lado si la experiencia es positiva, los pesos se fortalecen. Esto va a ser muy importante para un proceso (al final del curso) que se llama "backpropagation" (propagacion hacia atras). El tipo de comportamiento aca explicado se llama **Comportamiento Hebbiano**

3. En 1958 (el texto que mostre ayer es de 1961) Frank Rosenblatt, en el
Cornell Aeronautical Laboratory describio un modelo que se aproxima al modelo neurogico biologico, muchas mas profundo que el de McCulloh and Pitts.

Asuma que usted quiere tomar un viaje. Hay parametros sobre los cuales usted no tiene control.

Los datos $x_i$. 
1. Distancia desde mi casa.
2. modo de transporte: desde ir a pie hasta ir en avion (barco, otros).
3. clima del destino. Humedo, seco, calor, frio.
4. costo: tiquetes y alojamiento

Los pesos. Viene preguntas importantes.

Pesos: tiene un valor numerico $w_i$
1. Que tan dispuestos estamos de viajar a distancias largas. 
2. Nos da miedo subirnos a un avion, o a un barco, nos encanta.
3. No soporto el frio, o el calor, o la humedad.
4. No tengo plata, o no quiero gastar, o si quiero gastar.

Voy a hacer el siguinte calculo $\sum w_i x_i > u$ voy donde $u$ es un
umbral que tu decides, si $\sum w_i x_i \le u$. No voy. 
Por ejemplo pesos positivos a cosas que nos gustan y negativos a cosas que nos desagradan. 

El cerebro toma decisiones en base a este modelo simple pero con muchisimos mas datos, muchisimas mas conexiones (capas)  y el resultado es un "risa" o un "grito" o "no cruzar la calle".

## Modelo de McCulloh-Pitts:

En este modelo es (puramente logico) y pensamos en las operaciones binarias
"o", "y" (mas adelante la XO).

Los pesos se usan son 0, 1, -1. La salida es binaria (boolean) 1 (cierto)
0 (falso).

Lo que queremos es el proceso de suma ponderada.


\begin{eqnarray}
\left \{
    \begin{array}{cc}
    1   & \text{si  } \sum w_i x_i \ge u  \\
    0   & \text{de lo contrario} 
    \end{array}
    \right . 
\end{eqnarray}

## Funciones de Activacion: 
Las funciones de activacion son las que hacen que un proceso de redes neuronales no sean lineales. Es malo ser lineal? no. Si el proceso es lineal
lo que estamos haciendo es **regresion lineal**. Pero la regresion lineal es limitada. Y no podemos entonces aproximar funciones no lineales y se le destruye la hipotesis de universalidad a las redes neuronales. 

1. La identidad $I(x)=x$. Esta funcion "no hace nada" deja todo igual y produce simplente linealidad y por ende el proces es una regresion lineal.

2. Funcion de escala (step function , Heaviside function).


\begin{eqnarray}
H(t) = \left \{
    \begin{array}{cc}
    1  & \text{si } t > 0 \\
    0  & \text{si } t \le 0 
    \end{array}
    \right  . 
\end{eqnarray}


3. Funcion logistica.
Es importante por que es como una probabilidad. 

\begin{eqnarray}
S(t) = \frac{1}{1 + \exp(-t)}
\end{eqnarray}

\text{ReLU} (t) = \max \{0, t \} = \left \{
