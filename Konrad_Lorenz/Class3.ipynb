### Softmax:

\begin{eqnarray}
\sigma : \mathbb{R}^{K} &\to& [0,1]^K \\
(z_i) &\mapsto& \sigma(z_i) = \frac{\exp(z)}{\sum_{j=1}^K \exp(z_j)}
\end{eqnarray}

La mas simple **Una neurona de entrada, una oculta y una de salida**.

Como se propaga la informacion hacia adelante.

La entrada $x$, luego viene el parte lineal.

$z = x \theta$.
Notacion: Al $\theta$ lo llamamos $\theta^{(1)}$, por que es peso que sale de la capa 1.

$z^{(2)} = x \theta^{(1)}$.
Tocas activar este $z$. Usamos la funcion logistica.

$$ h(\theta, x) = a^{(2)} = S( z^{(2)}) = \frac{1}{1 + \exp(-z^{(2)})}$$
En Machine Learning $h(\theta, x)$ es la hipotesis, es aquello que se compara
con las etiquetas $y_i$ .




```python
import numpy as np
import matplotlib.pyplot as plt

def s(x):
    return 1/(1 + np.exp(-x))

dom = np.arange(-9,9,0.1)
plt.grid()
plt.plot(dom, s(dom))

s(-30)
```

Nodo amarillo. Se llama "bias" en espanol sesgo o (intercepto).

La funcion lineal simple es

$$ \theta_0 + \theta_1 x$$.  (note que $\theta_0 = 1 \theta_0$.
Esto lo podemos  escribir como un producto de matrices


\begin{eqnarray}
(1 \quad x)^T \begin{pmatrix} \theta_0 \\ \theta_1 \end{pmatrix}
\end{eqnarray}

En el caso de un hiperplano

\begin{eqnarray}
(1 \quad x_1 \quad \cdots \quad x_n) \begin{pmatrix} \theta_0  \\ \theta_1 \\ \vdots \\ \theta_n 
\end{pmatrix} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \cdots + \theta_n x_n
\end{eqnarray}

Si los datos de entrada son matriciales.
La operacion seria

\begin{eqnarray}
\begin{pmatrix}
1 & x_{11} & \cdots & x_{1n} \\
1 & x_{21} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
1 & x_{m1} & \cdots & x_{mn} 
\end{pmatrix}
\end{eqnarray}
